## 1. Variables y Cuantificadores

### 1.1 La Naturaleza de las Variables

En matemáticas y computación, el concepto de **variable** tiene diversas interpretaciones estructuradas. Siguiendo a Morales et al., *"una variable es un símbolo que representa una cantidad, relación u otras estructuras matemáticas"*.

Podemos clasificar la variabilidad en tres formas fundamentales:
1. **Variables que no varían (Incógnitas):** Tienen un valor fijo y único en un contexto dado, aunque desconocido *a priori*. Ejemplo: En la ecuación del lado de un triángulo equilátero con perímetro 21:
   $$3x = 21 \implies x = 7$$
2. **Variables Discretas:** Toman valores en un conjunto discreto (finito o infinito numerable). Ejemplo: El costo $M$ de fabricar $x$ objetos a un precio de $3$ unidades por objeto:
   $$M = 3x, \quad x \in \mathbb{N}$$
3. **Variables Continuas:** Toman cualquier valor dentro de un intervalo continuo de números reales. Ejemplo: La ecuación de la recta de posición en física con velocidad constante:
   $$y = 5x, \quad x \in \mathbb{R}$$

#### Concepciones y Usos Algebraicos (Usiskin, Juárez, Ursini)
Las variables toman distintos roles según la rama de las matemáticas:
* **Generalizadores de patrones:** En aritmética generalizada (ej. $a + b = b + a$).
* **Incógnitas/Constantes:** En procedimientos de resolución de ecuaciones.
* **Argumentos/Parámetros:** Al estudiar relaciones entre cantidades (funciones).
* **Marcas arbitrarias en el papel:** Al estudiar estructuras abstractas.

### 1.2 Funciones Proposicionales y Cuantificadores

Una **función proposicional** (o predicado) es una sentencia que contiene una o más variables y que se convierte en una proposición (con un valor de verdad definido: Verdadero o Falso) una vez que se le asignan valores específicos a dichas variables. Se denota por $P(x)$, donde $x$ pertenece a un dominio de discurso $D$.

#### Cuantificación Formal
Existen dos cuantificadores fundamentales para transformar funciones proposicionales en proposiciones sobre todo un dominio $D$:

1. **Cuantificador Universal ($\forall$):** La proposición $\forall x \in D, P(x)$ es verdadera si y solo si $P(x)$ es verdadera para **todos** los valores de $x$ en el dominio $D$.
2. **Cuantificador Existencial ($\exists$):** La proposición $\exists x \in D, P(x)$ es verdadera si y solo si $P(x)$ es verdadera para **al menos un** valor de $x$ en el dominio $D$.
3. **Cuantificador Existencial Único ($\exists!$):** Indica que existe **exactamente un** elemento en el dominio que cumple la propiedad.

#### Negación de Cuantificadores (Leyes de De Morgan Cuantificacionales)
La negación de una cuantificación sigue las siguientes equivalencias lógicas:
$$\neg (\forall x \in D, P(x)) \equiv \exists x \in D, \neg P(x)$$
$$\neg (\exists x \in D, P(x)) \equiv \forall x \in D, \neg P(x)$$

#### Anidación de Cuantificadores
Cuando trabajamos con múltiples variables, podemos anidar cuantificadores. El orden de los cuantificadores es crucial cuando son de distinto tipo:
* **Propiedad Conmutativa en $\mathbb{R}$:**
  $$\forall x \in \mathbb{R}, \forall y \in \mathbb{R} \ (x + y = y + x)$$
* **Inverso Aditivo en $\mathbb{R}$:**
  $$\forall x \in \mathbb{R}, \exists y \in \mathbb{R} \ (x + y = 0)$$
  *(Nota: Aquí el orden importa. Si dijéramos $\exists y \in \mathbb{R}, \forall x \in \mathbb{R} \ (x + y = 0)$, significaría que hay un único número $y$ que es el inverso de absolutamente todos los números reales a la vez, lo cual es falso).*


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo de los gráficos
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 1. Variable Discreta: Costo de fabricación M = 3x (x entero positivo)
x_discreto = np.arange(1, 11)
M_discreto = 3 * x_discreto

ax1.stem(x_discreto, M_discreto, linefmt='b-', markerfmt='bo', basefmt='r-')
ax1.set_title('Variable Discreta: Costo de Fabricación $M = 3x$', fontsize=13, fontweight='bold')
ax1.set_xlabel('Cantidad de Mesas ($x \in \mathbb{N}$)', fontsize=11)
ax1.set_ylabel('Costo Total ($M$)', fontsize=11)
ax1.set_xticks(x_discreto)
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Variable Continua: Posición en función del tiempo y = 5x (x real positivo)
x_continuo = np.linspace(0, 10, 200)
y_continuo = 5 * x_continuo

ax2.plot(x_continuo, y_continuo, 'r-', linewidth=2.5, label='$y = 5x$')
ax2.set_title('Variable Continua: Trayectoria Rectilínea $y = 5x$', fontsize=13, fontweight='bold')
ax2.set_xlabel('Tiempo ($x \in \mathbb{R}^+$)', fontsize=11)
ax2.set_ylabel('Posición ($y$)', fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Dominio finito de prueba (universo de discurso)
D = list(range(-10, 11)) # Enteros entre -10 y 10

# Definimos algunas funciones proposicionales
def P(x):
    # P(x): "x^2 >= 0"
    return x**2 >= 0

def Q(x):
    # Q(x): "x es par"
    return x % 2 == 0

def R(x):
    # R(x): "x + 1 > x"
    return x + 1 > x

def S(x):
    # S(x): "x > 5"
    return x > 5

def evaluar_cuantificadores(predicado, dominio, nombre_predicado):
    # Cuantificador Universal (all)
    universal = all(predicado(x) for x in dominio)
    # Cuantificador Existencial (any)
    existencial = any(predicado(x) for x in dominio)
    # Cuantificador Existencial Único
    existencial_unico = sum(1 for x in dominio if predicado(x)) == 1
    
    print(f"Predicado {nombre_predicado}:")
    print(f"  - Universal (\\forall x P(x)): {universal}")
    print(f"  - Existencial (\\exists x P(x)): {existencial}")
    print(f"  - Existencial Único (\\exists! x P(x)): {existencial_unico}")
    print("-" * 50)

evaluar_cuantificadores(P, D, "P(x): x^2 >= 0")
evaluar_cuantificadores(Q, D, "Q(x): x es par")
evaluar_cuantificadores(R, D, "R(x): x + 1 > x")
evaluar_cuantificadores(S, D, "S(x): x > 5")
```

## 2. Reglas de Inferencia y Demostración de Teoremas

Un **teorema** es una sentencia matemática cuya veracidad puede ser establecida mediante una **demostración** (una secuencia lógica de pasos fundamentados en axiomas, definiciones e hipótesis). Las **reglas de inferencia** son los patrones lógicos que permiten deducir de manera válida nuevas afirmaciones a partir de premisas conocidas.

### 2.1 Conceptos Fundamentales de la Lógica Proposicional
* **Tautología:** Una proposición compuesta que es siempre verdadera, independientemente de los valores de verdad de las proposiciones simples que la componen (ej. $p \lor \neg p$).
* **Contradicción:** Una proposición compuesta que es siempre falsa (ej. $p \land \neg p$).
* **Contingencia:** Una proposición compuesta que puede ser verdadera o falsa según los valores de sus componentes (ej. $p \to q$).

### 2.2 Principales Reglas de Inferencia

A continuación se detallan las reglas de inferencia proposicionales más importantes, estructuradas como implicaciones tautológicas:

| Nombre | Regla de Inferencia | Tautología Asociada | Descripción |
| :--- | :---: | :---: | :--- |
| **Adición** | $\frac{p}{\therefore p \lor q}$ | $p \to (p \lor q)$ | Si $p$ es verdadera, entonces su disyunción con cualquier otra proposición también lo es. |
| **Simplificación** | $\frac{p \land q}{\therefore p}$ | $(p \land q) \to p$ | Si una conjunción es verdadera, cada una de sus partes lo es. |
| **Conjunción** | $\frac{p, \ q}{\therefore p \land q}$ | $((p) \land (q)) \to (p \land q)$ | Si dos proposiciones son verdaderas por separado, su conjunción lo es. |
| **Modus Ponens** | $\frac{p, \ p \to q}{\therefore q}$ | $[p \land (p \to q)] \to q$ | "El modo que afirma". Si la hipótesis y la implicación son verdaderas, la conclusión es verdadera. |
| **Modus Tollens** | $\frac{\neg q, \ p \to q}{\therefore \neg p}$ | $[\neg q \land (p \to q)] \to \neg p$ | "El modo que niega". Si el consecuente es falso, el antecedente debe ser falso. |
| **Silogismo Hipotético** | $\frac{p \to q, \ q \to r}{\therefore p \to r}$ | $[(p \to q) \land (q \to r)] \to (p \to r)$ | Transitividad de la implicación. |
| **Silogismo Disyuntivo** | $\frac{p \lor q, \ \neg p}{\therefore q}$ | $[(p \lor q) \land \neg p] \to q$ | Si una de dos opciones es verdadera y se descarta una, la otra debe ser verdadera. |
| **Ley de Resolución** | $\frac{p \lor q, \ \neg p \lor r}{\therefore q \lor r}$ | $[(p \lor q) \land (\neg p \lor r)] \to (q \lor r)$ | Base para los algoritmos de resolución automática en Inteligencia Artificial. |

### 2.3 Reglas de Inferencia para Sentencias Cuantificadas

Cuando las premisas involucran cuantificadores, utilizamos reglas que nos permiten pasar de proposiciones cuantificadas a casos particulares (instanciación), operar con ellas, y volver a generalizar:

* **Particularización Universal (UI):**
  $$\frac{\forall x P(x)}{\therefore P(c) \quad \text{para un } c \text{ específico del dominio}}$$
  *Ejemplo:* "Todos los seres humanos son mortales. Sócrates es humano. Por lo tanto, Sócrates es mortal".
* **Generalización Universal (UG):**
  $$\frac{P(c) \quad \text{para un } c \text{ arbitrario}}{\therefore \forall x P(x)}$$
  *(Nota: El elemento $c$ debe ser genérico y no tener condiciones especiales).*
* **Particularización Existencial (EI):**
  $$\frac{\exists x P(x)}{\therefore P(c) \quad \text{para algún } c \text{ del dominio}}$$
* **Generalización Existencial (EG):**
  $$\frac{P(c) \quad \text{para un } c \text{ específico}}{\therefore \exists x P(x)}$$


```python
import sympy as sp
from sympy.logic.boolalg import truth_table

# Definimos las variables proposicionales
p, q, r = sp.symbols('p q r')

# 1. Comprobamos la Tautología del Modus Ponens: [p & (p -> q)] -> q
impl = p & (p >> q)
tautologia_mp = impl >> q

print("Tabla de verdad para el Modus Ponens: [p & (p -> q)] -> q")
# Generamos la tabla de verdad
variables = [p, q]
table = list(truth_table(tautologia_mp, variables))

header = f"{'p':<6} | {'q':<6} | {'[p & (p -> q)] -> q':<20}"
print(header)
print("-" * len(header))
for row in table:
    # row es ([p_val, q_val], result)
    inputs, result = row
    vals = [str(val) for val in inputs] + [str(result)]
    print(f"{vals[0]:<6} | {vals[1]:<6} | {vals[2]:<20}")
print("\n¿Es una tautología?", sp.simplify_logic(tautologia_mp) == True)
print("=" * 60)

# 2. Comprobamos la contradicción p & ~p
contradiccion = p & ~p
print("\nTabla de verdad para la contradicción: p & ~p")
variables_c = [p]
table_c = list(truth_table(contradiccion, variables_c))
header_c = f"{'p':<6} | {'p & ~p':<10}"
print(header_c)
print("-" * len(header_c))
for row in table_c:
    inputs, result = row
    vals = [str(val) for val in inputs] + [str(result)]
    print(f"{vals[0]:<6} | {vals[1]:<10}")
print("\n¿Es una contradicción?", sp.simplify_logic(contradiccion) == False)
```

### 2.4 Ejemplo de Aplicación de Reglas de Inferencia

Consideremos el siguiente argumento clásico expuesto en el PDF. Queremos deducir la conclusión **"estaremos en casa para la puesta del sol"** a partir de las siguientes hipótesis:
1. *"Esta tarde no hace sol y hace más frío que ayer."*
2. *"Iremos a nadar sólo si hace sol."*
3. *"Si no vamos a nadar, daremos un paseo en canoa."*
4. *"Si damos un paseo en canoa, estaremos en casa para la puesta del sol."*

#### Formalización Lógica:
Definimos las proposiciones atómicas:
* $p$: "Esta tarde hace sol"
* $q$: "Hace más frío que ayer"
* $r$: "Iremos a nadar"
* $s$: "Daremos un paseo en canoa"
* $t$: "Estaremos en casa para la puesta del sol"

Las hipótesis se traducen formalmente como:
1. $\neg p \land q$
2. $r \to p$
3. $\neg r \to s$
4. $s \to t$

#### Demostración Paso a Paso:

| Paso | Expresión Lógica | Regla Utilizada / Justificación |
| :---: | :---: | :--- |
| **1** | $\neg p \land q$ | Hipótesis 1 |
| **2** | $\neg p$ | Simplificación aplicada al Paso 1 |
| **3** | $r \to p$ | Hipótesis 2 |
| **4** | $\neg r$ | Modus Tollens aplicado a los Pasos 2 y 3 |
| **5** | $\neg r \to s$ | Hipótesis 3 |
| **6** | $s$ | Modus Ponens aplicado a los Pasos 4 y 5 |
| **7** | $s \to t$ | Hipótesis 4 |
| **8** | $t$ | **Modus Ponens aplicado a los Pasos 6 y 7 (Conclusión)** |


```python
import sympy as sp

# Definimos los símbolos
p, q, r, s, t = sp.symbols('p q r s t')

# Definimos las hipótesis
h1 = ~p & q       # Esta tarde no hace sol y hace más frío que ayer
h2 = r >> p       # Iremos a nadar sólo si hace sol (r -> p)
h3 = ~r >> s      # Si no vamos a nadar, daremos un paseo en canoa (~r -> s)
h4 = s >> t       # Si damos un paseo en canoa, estaremos en casa para la puesta del sol (s -> t)

# La conjunción de todas las hipótesis
hipotesis = h1 & h2 & h3 & h4

# Queremos verificar si la conclusión 't' se deduce lógicamente de las hipótesis.
# Esto es equivalente a comprobar si la implicación (hipotesis -> t) es una TAUTOLOGÍA.
teorema = hipotesis >> t

# Simplificamos la lógica del teorema
es_tautologia = sp.simplify_logic(teorema)

print("Demostración computacional de la validez del argumento:")
print(f"Hipótesis unificadas: {sp.pretty(hipotesis)}")
print(f"Teorema a demostrar: (Hipótesis) -> t")
print(f"Resultado de la simplificación: {es_tautologia}")

if es_tautologia == True:
    print("\n¡ÉXITO! La conclusión 't' (estaremos en casa para la puesta del sol) es LÓGICAMENTE VÁLIDA.")
else:
    print("\nERROR: El argumento no es válido.")
```

## 3. Métodos de Demostración de Teoremas

En esta sección estudiaremos las técnicas sistemáticas utilizadas para demostrar la veracidad de sentencias condicionales de la forma $p \to q$.

### 3.1 Demostración Directa
Consiste en asumir que la hipótesis $p$ es **verdadera** y, utilizando axiomas, definiciones y teoremas previos, deducir lógicamente que la tesis $q$ también debe ser **verdadera**.

### 3.2 Demostración Indirecta (Por Contraposición)
Se basa en la equivalencia lógica entre una implicación y su contrapositiva:
$$(p \to q) \equiv (\neg q \to \neg p)$$
Asumimos que la tesis es falsa ($\neg q$) y demostramos que, bajo esa premisa, la hipótesis también debe ser falsa ($\neg p$).

#### Ejemplo de Demostración Indirecta:
* **Teorema:** Si $n$ es un entero y $n^2$ es impar, entonces $n$ es impar.
* **Demostración por Contraposición:**
  * Asumimos la negación de la tesis: $n$ **no es impar** (es decir, $n$ es par).
  * Por definición de número par: $n = 2k$ para algún entero $k$.
  * Elevamos al cuadrado ambos lados:
    $$n^2 = (2k)^2 = 4k^2 = 2(2k^2)$$
  * Como $2k^2$ es un entero (llamémoslo $m$), tenemos $n^2 = 2m$, lo que implica que $n^2$ **es par** (no es impar).
  * Esto demuestra $\neg q \to \neg p$. Por lo tanto, el teorema original $p \to q$ es verdadero.

### 3.3 Demostración por Reducción al Absurdo (Contradicción)
Para demostrar que una proposición $p$ es verdadera, asumimos temporalmente que es **falsa** ($\neg p$) y derivamos una contradicción de la forma $r \land \neg r$ (un absurdo lógico). Dado que las contradicciones son siempre falsas, la suposición original $\neg p$ debe ser falsa, lo que obliga a que $p$ sea verdadera.
$$\neg p \to F \implies p \text{ es verdadera}$$

#### Ejemplo Clásico: Irracionalidad de $\sqrt{2}$
* **Teorema:** $\sqrt{2}$ es irracional.
* **Demostración por Contradicción:**
  * Asumimos la negación: $\sqrt{2}$ **es racional**.
  * Por definición, existen enteros $a$ y $b$ ($b \neq 0$) sin factores comunes tales que:
    $$\sqrt{2} = \frac{a}{b}$$
  * Elevamos al cuadrado y despejamos:
    $$2 = \frac{a^2}{b^2} \implies a^2 = 2b^2$$
  * Esto significa que $a^2$ es par, lo cual implica que $a$ es par (se puede demostrar usando contraposición). Por tanto, $a = 2c$ para algún entero $c$.
  * Sustituimos $a$:
    $$(2c)^2 = 2b^2 \implies 4c^2 = 2b^2 \implies b^2 = 2c^2$$
  * Esto implica que $b^2$ es par, y por ende $b$ también es par.
  * Llegamos a una contradicción: $a$ y $b$ son ambos pares (tienen factor común 2), lo cual contradice nuestra premisa de que la fracción $\frac{a}{b}$ era irreducible.
  * Por lo tanto, la suposición de que $\sqrt{2}$ es racional es falsa, y $\sqrt{2}$ es irracional.

### 3.4 Demostración por Casos
Si la hipótesis es una disyunción de la forma $p_1 \lor p_2 \lor \dots \lor p_n$, demostramos el teorema probando de forma individual cada implicación $p_i \to q$:
$$[(p_1 \lor p_2 \lor \dots \lor p_n) \to q] \equiv [(p_1 \to q) \land (p_2 \to q) \land \dots \land (p_n \to q)]$$

### 3.5 Demostración de Existencia y Unicidad
* **Existencia ($\exists x P(x)$):**
  * **Constructiva:** Encontrar o construir explícitamente un elemento $c$ tal que $P(c)$ sea verdadero.
  * **No constructiva:** Probar que debe existir tal elemento sin mostrarlo (frecuentemente por contradicción).
* **Unicidad ($\exists! x P(x)$):** Se demuestra en dos partes:
  1. *Existencia:* Se prueba que existe al menos un elemento $x$ tal que $P(x)$.
  2. *Unicidad:* Se asume que existen dos elementos $x$ e $y$ que cumplen la propiedad y se demuestra que necesariamente $x = y$.

### 3.6 Demostración por Contraejemplo
Para refutar una afirmación universal de la forma $\forall x P(x)$ (es decir, demostrar que es falsa), basta con encontrar **un solo elemento** $c$ en el dominio para el cual $P(c)$ sea falso.
$$\neg (\forall x P(x)) \equiv \exists x \neg P(x)$$


```python
# Teorema propuesto: "Todo entero positivo es la suma de los cuadrados de tres enteros"
# Queremos buscar si existe un contraejemplo para esta afirmación.
# Es decir, buscar un entero positivo n tal que n != a^2 + b^2 + c^2 para todo a, b, c enteros.

def es_suma_de_tres_cuadrados(n):
    # Dado que los cuadrados no pueden exceder n, limitamos la búsqueda
    limite = int(n**0.5) + 1
    for a in range(limite):
        for b in range(limite):
            for c in range(limite):
                if a**2 + b**2 + c**2 == n:
                    return True, (a, b, c)
    return False, None

# Probamos para los primeros enteros positivos
max_n = 20
print(f"Evaluando la propiedad para los enteros del 1 al {max_n}:\\n")

contraejemplo_encontrado = False

for n in range(1, max_n + 1):
    valido, combinacion = es_suma_de_tres_cuadrados(n)
    if valido:
        print(f"n = {n:<3} -> VÁLIDO: {combinacion[0]}^2 + {combinacion[1]}^2 + {combinacion[2]}^2 = {n}")
    else:
        print(f"n = {n:<3} -> ¡CONTRAEJEMPLO DETECTADO! No se puede escribir como suma de 3 cuadrados.")
        contraejemplo_encontrado = True
        # El primer contraejemplo es suficiente para refutar la afirmación universal.
        break

if contraejemplo_encontrado:
    print("\\nConclusión: La afirmación universal ha sido refutada de forma computacional mediante contraejemplo.")
```

## 4. Inducción Matemática

El principio de **inducción matemática** es una técnica fundamental para demostrar afirmaciones de la forma $\forall n \in \mathbb{N}, P(n)$ (proposiciones indexadas sobre los números enteros positivos o naturales).

### 4.1 Principio de Inducción Completa (Simple)

Para probar que $P(n)$ es verdadera para todo entero positivo $n$, realizamos dos pasos:
1. **Caso Base:** Demostrar que la proposición es verdadera para el primer elemento del dominio (usualmente $n = 1$, es decir, $P(1)$ es verdadero).
2. **Paso Inductivo:** Demostrar que **si** la proposición se asume verdadera para un entero genérico $k$ (llamada **Hipótesis Inductiva**), **entonces** también es verdadera para el siguiente entero $k + 1$.
   $$P(k) \implies P(k+1)$$

Si se cumplen ambos pasos, entonces $P(n)$ es verdadera para todo $n \in \mathbb{N}$.

#### Demostración Analítica de Ejemplo:
* **Teorema:** Demostrar que para todo $n \in \mathbb{N}$, se cumple que $n < 2^n$.
* **Demostración por Inducción:**
  * **Caso Base ($n = 1$):**
    $$1 < 2^1 \implies 1 < 2 \quad (\text{Verdadero})$$
  * **Paso Inductivo:**
    * Asumimos la **Hipótesis Inductiva**: $k < 2^k$ para un entero $k \ge 1$.
    * Debemos demostrar que: $k + 1 < 2^{k+1}$.
    * Partimos de la hipótesis inductiva sumando $1$ a ambos lados de la desigualdad:
      $$k + 1 < 2^k + 1$$
    * Dado que para todo $k \ge 1$ se cumple la desigualdad obvia $1 \le 2^k$, podemos escribir:
      $$2^k + 1 \le 2^k + 2^k = 2 \cdot 2^k = 2^{k+1}$$
    * Por transitividad:
      $$k + 1 < 2^{k+1}$$
    * El paso inductivo se cumple. Por lo tanto, $n < 2^n$ para todo $n \in \mathbb{N}$.

### 4.2 Principio de Inducción Fuerte

En algunos casos, la hipótesis inductiva de que la propiedad se cumple para el paso inmediatamente anterior ($k$) no es suficiente para demostrar el paso $k+1$. En su lugar, se asume que la propiedad es verdadera para **todos los enteros desde el caso base hasta $k$**.

**Pasos de la Inducción Fuerte:**
1. **Caso Base:** Demostrar $P(1)$ (o varios casos base si es necesario).
2. **Paso Inductivo:** Demostrar que si $[P(1) \land P(2) \land \dots \land P(k)]$ es verdadero, entonces $P(k+1)$ es verdadero.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Graficamos la comparación de n vs 2^n para ver el crecimiento
n_valores = np.arange(1, 11)
linear = n_valores
exponential = 2**n_valores

plt.figure(figsize=(10, 5))
plt.plot(n_valores, linear, 'o--', label='Fórmula Lineal: $n$')
plt.plot(n_valores, exponential, 's-', label='Crecimiento Exponencial: $2^n$')
plt.yscale('log') # Escala logarítmica para apreciar la diferencia de órdenes
plt.title('Demostración Gráfica de $n < 2^n$ (Escala Logarítmica)', fontsize=13, fontweight='bold')
plt.xlabel('Entero $n$', fontsize=11)
plt.ylabel('Valor (Escala Log)', fontsize=11)
plt.xticks(n_valores)
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

# 2. Verificación computacional del paso inductivo
def verificar_paso_inductivo(k):
    # Hipótesis Inductiva: asumimos que k < 2^k es verdadero
    hipotesis = k < 2**k
    
    # Evaluamos el paso k+1
    consecuente = (k + 1) < 2**(k + 1)
    
    # La implicación lógica Hipótesis -> Consecuente es equivalente a (~Hipotesis | Consecuente)
    implicacion = (not hipotesis) or consecuente
    return implicacion

# Validamos para un rango de valores de k
rango_k = range(1, 100)
todos_validos = all(verificar_paso_inductivo(k) for k in rango_k)
print(f"¿Es válido el paso inductivo P(k) -> P(k+1) para todo k en 1..100? {todos_validos}")
```

## 5. Recursión

La **recursión** (o recursividad) es un método para definir un objeto, función o proceso en términos de sí mismo. En programación y matemáticas, una definición o función recursiva consta de:
1. **Paso Base (o Caso Base):** Especifica el valor o comportamiento de la función en uno o más puntos iniciales (ej. 0 o 1). Este paso detiene la recursión y evita bucles infinitos.
2. **Paso Recursivo:** Proporciona una regla para calcular el valor de la función en un punto utilizando valores de la misma función en puntos más pequeños.

### 5.1 Ejemplo Clásico: La Función Factorial ($n!$)

El factorial de un número entero no negativo $n$ se denota por $n!$ y representa el producto de todos los enteros positivos menores o iguales a $n$.

* **Definición Recursiva:**
  * **Caso Base:**
    $$0! = 1$$
  * **Paso Recursivo:**
    $$n! = n \cdot (n - 1)! \quad \text{para } n \ge 1$$

#### Traza de Ejecución de $7!$:
$$7! = 7 \cdot 6!$$
$$7! = 7 \cdot (6 \cdot 5!)$$
$$7! = 7 \cdot 6 \cdot 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1 \cdot 0!$$
$$7! = 7 \cdot 6 \cdot 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1 \cdot 1 = 5040$$


```python
import sys

# Aumentamos el límite de recursión por si acaso (por defecto suele ser 1000)
sys.setrecursionlimit(2000)

def factorial_con_traza(n, profundidad=0):
    # Formateo visual de la indentación según la profundidad de la pila de llamadas
    indentacion = "    " * profundidad
    print(f"{indentacion}--> Llamando a factorial({n})")
    
    # 1. Caso Base
    if n == 0:
        print(f"{indentacion}<-- Caso Base alcanzado: factorial(0) = 1")
        return 1
    
    # 2. Paso Recursivo
    resultado_parcial = factorial_con_traza(n - 1, profundidad + 1)
    resultado = n * resultado_parcial
    
    print(f"{indentacion}<-- Retornando: factorial({n}) = {n} * {resultado_parcial} = {resultado}")
    return resultado

# Ejecutamos el factorial de 7 con visualización de la pila de llamadas (call stack)
n_prueba = 7
print(f"Visualización del Call Stack para factorial({n_prueba}):")
print("=" * 60)
resultado_final = factorial_con_traza(n_prueba)
print("=" * 60)
print(f"Resultado final: {n_prueba}! = {resultado_final}")
```

## 6. Resumen de la Lección

1. **Variables:** Son abstracciones simbólicas cuyos usos van desde la representación de incógnitas algebraicas fijas hasta variables continuas o discretas utilizadas para modelar fenómenos físicos y computacionales.
2. **Cuantificadores:** Permiten formalizar sentencias universales ($\forall$) y existenciales ($\exists$), estableciendo una base rigurosa para la lógica de predicados.
3. **Inferencia Lógica:** Utiliza reglas sintácticas tautológicas (como el Modus Ponens y Modus Tollens) para deducir de forma segura verdades nuevas a partir de hipótesis validadas.
4. **Métodos de Demostración:** Proporcionan herramientas analíticas (demostración directa, por contraposición, reducción al absurdo y por contraejemplo) para demostrar la validez de teoremas.
5. **Inducción:** Es la herramienta matemática por excelencia para probar propiedades sobre conjuntos infinitos estructurados ordenadamente (como los números naturales).
6. **Recursión:** Traduce la definición inductiva a algoritmos informáticos eficientes, requiriendo siempre un caso base bien definido para evitar la recursión infinita.

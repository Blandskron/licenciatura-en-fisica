## 1. Fundamentos de la Teoría de Conjuntos

El concepto de **conjunto** es una de las ideas unificadoras de las matemáticas. De acuerdo con Kenneth [1], *"un conjunto es una colección desordenada de objetos"*. Intuitivamente, un conjunto está formado de acuerdo con una regla que permite decidir si un objeto dado pertenece o no a dicha colección.

### 1.1 Elementos y Pertenencia
* Si un objeto $a$ forma parte de un conjunto $A$, se dice que $a$ **pertenece** a $A$ y se escribe $a \in A$.
* Si no forma parte de él, se escribe $a \notin A$.
* Por convención, se utilizan letras mayúsculas ($A, B, C, \dots$) para los conjuntos y letras minúsculas ($a, b, x, y, \dots$) para sus elementos.

### 1.2 Definición de Conjuntos
Existen dos formas fundamentales para definir y describir un conjunto:
1. **Por Extensión (o Enumeración):** Se listan explícitamente todos los elementos que componen el conjunto entre llaves, separados por comas.
   $$A = \{0, 1\}$$
   $$B = \{\text{Verdadero}, \text{Falso}\}$$
2. **Por Comprensión (o Condición):** Se establece la propiedad o propiedad(es) lógica(s) que deben cumplir todos los elementos del conjunto. Su formato general es:
   $$E = \{x \mid P(x)\} \quad \text{o} \quad E = \{x : P(x)\}$$
   Donde la barra vertical ($\mid$) o los dos puntos ($:$) se leen *"tal que"* o *"tales que"*, y $P(x)$ es la función proposicional que actúa como filtro.
   *Ejemplo:* El conjunto de números primos mayores a 1 se define como $F = \{x \in \mathbb{N} \mid x \text{ es número primo}\}$.

### 1.3 Subconjuntos, Conjunto Universal y Conjunto Vacío
* **Igualdad de Conjuntos ($A = B$):** Dos conjuntos son iguales si y solo si tienen exactamente los mismos elementos. El orden de escritura y la repetición de elementos no alteran al conjunto:
  $$\{1, 2, 3\} = \{3, 1, 2\} = \{1, 2, 1, 3, 2, 1\}$$
* **Subconjunto ($B \subseteq A$):** Se dice que $B$ es un subconjunto de $A$ si todo elemento de $B$ pertenece también a $A$.
  $$B \subseteq A \iff \forall x (x \in B \implies x \in A)$$
* **Subconjunto Propio ($B \subset A$):** Se dice que $B$ es un subconjunto propio de $A$ si $B$ es subconjunto de $A$ pero es distinto de $A$ ($B \subseteq A$ y $B \neq A$).
* **Conjunto Universo ($U$):** El conjunto que contiene todos los posibles objetos de interés para un contexto determinado.
* **Conjunto Vacío ($\emptyset$):** El conjunto que carece de elementos.
  * **Teorema:** Para cualquier conjunto $A$, se cumple que $\emptyset \subseteq A$.
  * *Demostración por reducción al absurdo:* Supongamos que $\emptyset \not\subseteq A$. Por definición de subconjunto, esto implica que existe al menos un elemento $x$ tal que $x \in \emptyset$ y $x \notin A$. Sin embargo, esto contradice la propia definición del conjunto vacío, el cual no contiene ningún elemento. Por ende, la suposición original es falsa y $\emptyset \subseteq A$ es verdadero.

### 1.4 Sistemas Numéricos
En matemáticas, existen conjuntos de números con propiedades especiales estructurados por relaciones de inclusión sucesivas:
$$\mathbb{N} \subset \mathbb{Z} \subset \mathbb{Q} \subset \mathbb{R} \subset \mathbb{C}$$
* **Naturales ($\mathbb{N}$):** Números para contar $\{1, 2, 3, \dots\}$ (a veces incluye al $0$).
* **Enteros ($\mathbb{Z}$):** Naturales, sus negativos y el cero.
* **Racionales ($\mathbb{Q}$):** Fracciones de la forma $a/b$ donde $a, b \in \mathbb{Z}$ y $b \neq 0$.
* **Reales ($\mathbb{R}$):** Límite continuo de los racionales (incluye irracionales como $\pi, \sqrt{2}$).
* **Imaginarios ($\mathbb{I}$):** Números de la forma $bi$ donde $b \in \mathbb{R}$ e $i = \sqrt{-1}$.
* **Complejos ($\mathbb{C}$):** Números de la forma $a + bi$ con $a, b \in \mathbb{R}$.


```python
# Definición de conjuntos por extensión
A = {1, 2, 3}
B = {3, 1, 2}
C = {1, 2, 1, 3, 2, 1}

# 1. Verificación de Igualdad
print("¿A es igual a B?", A == B)
print("¿A es igual a C?", A == C)

# 2. Relación de Pertenencia
elemento = 3
print(f"¿{elemento} pertenece a A? {elemento in A}")
print(f"¿8 pertenece a A? {8 in A}")

# 3. Relación de Inclusión (Subconjuntos)
sub_A = {1, 2}
print(f"¿{sub_A} es subconjunto de A? {sub_A.issubset(A)}")
print(f"¿A es subconjunto de A? {A.issubset(A)}")

# 4. Propiedades del Conjunto Vacío (set() en Python)
vacio = set()
print("¿El conjunto vacío es subconjunto de A?", vacio.issubset(A))
print("¿El conjunto vacío es igual a A?", vacio == A)

```

## 2. Operaciones con Conjuntos

A partir de conjuntos existentes, podemos construir nuevos conjuntos aplicando operaciones algebraicas fundamentales.

### 2.1 Definición de Operaciones

Sean $A$ y $B$ dos conjuntos en un universo $U$:

1. **Unión ($A \cup B$):** Conjunto de elementos que pertenecen a $A$ o a $B$ (o a ambos).
   $$A \cup B = \{x \in U \mid x \in A \lor x \in B\}$$
2. **Intersección ($A \cap B$):** Conjunto de elementos que pertenecen tanto a $A$ como a $B$.
   $$A \cap B = \{x \in U \mid x \in A \land x \in B\}$$
3. **Complemento ($A^c$ o $\overline{A}$):** Conjunto de elementos del universo $U$ que no pertenecen a $A$.
   $$A^c = \{x \in U \mid x \notin A\}$$
4. **Diferencia ($A - B$):** Conjunto de elementos que pertenecen a $A$ pero no a $B$.
   $$A - B = \{x \in U \mid x \in A \land x \notin B\}$$
5. **Diferencia Simétrica ($A \Delta B$):** Conjunto de elementos que pertenecen a $A$ o a $B$, pero no a ambos en simultaneidad.
   $$A \Delta B = (A - B) \cup (B - A) = \{x \in U \mid x \in A \oplus x \in B\}$$
6. **Conjunto de las Partes (Power Set - $\mathcal{P}(S)$):** El conjunto que contiene a todos los posibles subconjuntos de $S$.
   * **Teorema:** Si $S$ es un conjunto finito con cardinal $|S| = n$, el conjunto de las partes $\mathcal{P}(S)$ tiene cardinal:
     $$|\mathcal{P}(S)| = 2^n$$
7. **Producto Cartesiano ($A \times B$):** Conjunto de todos los pares ordenados $(a, b)$ tales que $a \in A$ y $b \in B$.
   $$A \times B = \{(a, b) \mid a \in A \land b \in B\}$$
   *(Nota: $(a, b) \neq (b, a)$ si $a \neq b$, por lo que $A \times B \neq B \times A$).*

### 2.2 Leyes de De Morgan
Las leyes de De Morgan establecen la dualidad entre la unión y la intersección a través del complemento:
1. $(A \cup B)^c = A^c \cap B^c$
2. $(A \cap B)^c = A^c \cup B^c$

### 2.3 Identidades Algebraicas entre Conjuntos

Las operaciones de conjuntos cumplen las siguientes leyes fundamentales:

| Ley | Unión ($U$) | Intersección ($n$) |
| :--- | :---: | :---: |
| **Identidad** | $A \cup \emptyset = A$ | $A \cap U = A$ |
| **Dominación** | $A \cup U = U$ | $A \cap \emptyset = \emptyset$ |
| **Idempotencia** | $A \cup A = A$ | $A \cap A = A$ |
| **Doble Complemento** | $\overline{(\overline{A})} = A$ | - |
| **Conmutativa** | $A \cup B = B \cup A$ | $A \cap B = B \cap A$ |
| **Asociativa** | $A \cup (B \cup C) = (A \cup B) \cup C$ | $A \cap (B \cap C) = (A \cap B) \cap C$ |
| **Distributiva** | $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$ | $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$ |
| **Absorción** | $A \cup (A \cap B) = A$ | $A \cap (A \cup B) = A$ |
| **Complemento** | $A \cup A^c = U$ | $A \cap A^c = \emptyset$ |


```python
import itertools

# Definimos el Universo y dos conjuntos de prueba
U = set(range(1, 11))  # U = {1, 2, ..., 10}
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# 1. Operaciones Básicas
print("Union (A U B):", A.union(B))
print("Interseccion (A n B):", A.intersection(B))
print("Diferencia (A - B):", A.difference(B))
print("Diferencia Simetrica (A Delta B):", A.symmetric_difference(B))

# Complemento de A (U - A)
A_c = U - A
B_c = U - B
print("Complemento de A (A^c):", A_c)

# 2. Verificación de la Primera Ley de De Morgan: (A U B)^c == A^c n B^c
izq = U - (A.union(B))
der = A_c.intersection(B_c)
print("\n¿Se cumple la 1ra Ley de De Morgan?", izq == der)

# 3. Producto Cartesiano A x B (para subconjuntos pequeños)
X = {'a', 'b'}
Y = {1, 2, 3}
prod_cartesiano = set(itertools.product(X, Y))
print(f"\nProducto Cartesiano X x Y ({len(prod_cartesiano)} elementos):")
print(prod_cartesiano)

# 4. Conjunto de las Partes P(S)
def conjunto_partes(S):
    lista_S = list(S)
    partes = []
    for r in range(len(lista_S) + 1):
        for comb in itertools.combinations(lista_S, r):
            partes.append(set(comb))
    return partes

S = {1, 2, 3}
P_S = conjunto_partes(S)
print(f"\nConjunto de las Partes de {S} ({len(P_S)} subconjuntos):")
for idx, sub in enumerate(P_S):
    print(f"  Subconjunto {idx}: {sub}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos una función para graficar diagramas de Venn sombreados de forma personalizada
def graficar_diagrama_venn(ax, operacion, titulo):
    # Generamos una malla densa de coordenadas para evaluar las ecuaciones de los círculos
    x = np.linspace(-2.2, 2.2, 500)
    y = np.linspace(-2.2, 2.2, 500)
    X, Y = np.meshgrid(x, y)
    
    # Círculo A: Centro en (-0.5, 0), Radio = 1.0
    inside_A = (X + 0.5)**2 + Y**2 <= 1.0
    # Círculo B: Centro en (0.5, 0), Radio = 1.0
    inside_B = (X - 0.5)**2 + Y**2 <= 1.0
    
    if operacion == "union":
        region = inside_A | inside_B
    elif operacion == "interseccion":
        region = inside_A & inside_B
    elif operacion == "diferencia":
        region = inside_A & ~inside_B
    elif operacion == "simetrica":
        region = inside_A ^ inside_B
    else:
        region = np.zeros_like(X)

    # Rellenamos la región de la operación seleccionada
    ax.contourf(X, Y, region, levels=[0.5, 1.5], colors=['#3b82f6'], alpha=0.6)
    
    # Dibujamos las líneas de los bordes de los círculos
    theta = np.linspace(0, 2*np.pi, 200)
    # Contorno A
    ax.plot(1.0*np.cos(theta) - 0.5, 1.0*np.sin(theta), 'k-', linewidth=2)
    # Contorno B
    ax.plot(1.0*np.cos(theta) + 0.5, 1.0*np.sin(theta), 'k-', linewidth=2)
    
    # Etiquetas de los conjuntos
    ax.text(-0.9, 1.2, "A", fontsize=14, fontweight='bold', color='black')
    ax.text(0.8, 1.2, "B", fontsize=14, fontweight='bold', color='black')
    
    # Configuración de los ejes
    ax.set_xlim(-2.2, 2.2)
    ax.set_ylim(-1.8, 1.8)
    ax.set_aspect('equal')
    ax.set_title(titulo, fontsize=13, fontweight='bold')
    ax.axis('off')

# Creamos una figura de 2x2 para las 4 operaciones fundamentales
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

graficar_diagrama_venn(axes[0, 0], "union", "Unión: $A \\cup B$")
graficar_diagrama_venn(axes[0, 1], "interseccion", "Intersección: $A \\cap B$")
graficar_diagrama_venn(axes[1, 0], "diferencia", "Diferencia: $A - B$")
graficar_diagrama_venn(axes[1, 1], "simetrica", "Diferencia Simétrica: $A \\Delta B$")

plt.tight_layout()
plt.show()

```

## 3. Funciones

El concepto de **función** formaliza la regla de asignación entre dos conjuntos de variables.

### 3.1 Definición Matemática

Sean $A$ y $B$ dos conjuntos. Una **función** $f$ de $A$ en $B$ (escrito $f : A \to B$) es una asignación que asocia a **exactamente un** elemento de $B$ con cada elemento de $A$.

* Si $f(a) = b$, se dice que:
  * $b$ es la **imagen** de $a$.
  * $a$ es la **preimagen** de $b$.
* **Dominio:** El conjunto de partida $A$.
* **Codominio:** El conjunto de llegada $B$.
* **Rango (o Imagen de $f$):** El conjunto de todos los elementos de $B$ que tienen preimagen en $A$:
  $$\text{Rango}(f) = \{b \in B \mid \exists a \in A \ (f(a) = b)\}$$
  *(Nota: El rango es un subconjunto del codominio, y no necesariamente es igual a él).*

#### Álgebra de Funciones Reales
Si $f_1$ y $f_2$ son funciones de $\mathbb{R}$ en $\mathbb{R}$, podemos sumarlas o multiplicarlas:
* Suma: $(f_1 + f_2)(x) = f_1(x) + f_2(x)$
* Producto: $(f_1 \cdot f_2)(x) = f_1(x) \cdot f_2(x)$

### 3.2 Clasificación de Funciones

Una función $f: A \to B$ puede presentar propiedades estructurales de inyectividad y sobreyectividad:

1. **Inyectiva (Uno a Uno):** Si elementos distintos del dominio tienen imágenes distintas.
   $$\forall x, y \in A \ (f(x) = f(y) \implies x = y)$$
   Equivalente a su contrapositiva: si $x \neq y$, entonces $f(x) \neq f(y)$.
2. **Sobreyectiva (Sobre o Suryectiva):** Si todos los elementos del codominio tienen al menos una preimagen en el dominio. Es decir, el rango es exactamente igual al codominio.
   $$\forall b \in B \ (\exists a \in A \mid f(a) = b)$$
3. **Biyectiva:** Si la función es **inyectiva y sobreyectiva a la vez**. Las funciones biyectivas establecen una correspondencia uno a uno perfecta entre todos los elementos de $A$ y $B$.

### 3.3 Función Inversa y Composición

* **Función Inversa ($f^{-1}$):** Si $f: A \to B$ es una función **biyectiva**, entonces existe su función inversa $f^{-1}: B \to A$, la cual revierte la asignación:
  $$f^{-1}(b) = a \iff f(a) = b$$
* **Composición de Funciones ($f \circ g$):** Sean $g: A \to B$ y $f: B \to C$. La composición $f \circ g$ es una función de $A$ en $C$ definida por:
  $$(f \circ g)(a) = f(g(a))$$
  *(Nota: Para poder realizar la composición, la imagen de la función interna $g$ debe estar contenida en el dominio de la función externa $f$).*


```python
class RelacionFinita:
    def __init__(self, dominio, codominio, asignaciones):
        # asignaciones es un diccionario {a: b}
        self.A = set(dominio)
        self.B = set(codominio)
        self.f = asignaciones

    def es_funcion_valida(self):
        # 1. Cada elemento del dominio debe tener exactamente una asignación
        for a in self.A:
            if a not in self.f:
                return False
        # 2. Las imágenes deben estar en el codominio B
        for b in self.f.values():
            if b not in self.B:
                return False
        return True

    def analizar(self):
        if not self.es_funcion_valida():
            print("No es una función válida de A en B.")
            return

        # Comprobamos Inyectividad: si f(x) == f(y) => x == y
        imagenes = list(self.f.values())
        inyectiva = len(imagenes) == len(set(imagenes))

        # Comprobamos Sobreyectividad: si Rango(f) == Codominio B
        rango = set(imagenes)
        sobreyectiva = rango == self.B

        # Biyectiva
        biyectiva = inyectiva and sobreyectiva

        print("Análisis de la Función:")
        print(f"  - Inyectiva: {inyectiva}")
        print(f"  - Sobreyectiva: {sobreyectiva}")
        print(f"  - Biyectiva: {biyectiva}")
        print(f"  - Rango de f: {rango}")

# Ejemplos para probar (del diagrama de la Figura 10 del PDF)
print("--- Caso 1 (Inyectiva pero no sobreyectiva) ---")
# A = {1, 2, 3}, B = {A, B, C, D}
# f(1)=D, f(2)=B, f(3)=C
f1 = RelacionFinita({1, 2, 3}, {'A', 'B', 'C', 'D'}, {1: 'D', 2: 'B', 3: 'C'})
f1.analizar()

print("\n--- Caso 2 (Sobreyectiva pero no inyectiva) ---")
# A = {1, 2, 3, 4}, B = {A, B, C}
# f(1)=D, f(2)=B, f(3)=C, f(4)=D (Suponiendo codominio {'D', 'B', 'C'})
f2 = RelacionFinita({1, 2, 3, 4}, {'D', 'B', 'C'}, {1: 'D', 2: 'B', 3: 'C', 4: 'D'})
f2.analizar()

print("\n--- Caso 3 (Biyectiva) ---")
# A = {1, 2, 3, 4}, B = {A, B, C, D}
# f(1)=D, f(2)=B, f(3)=C, f(4)=A
f3 = RelacionFinita({1, 2, 3, 4}, {'A', 'B', 'C', 'D'}, {1: 'D', 2: 'B', 3: 'C', 4: 'A'})
f3.analizar()

print("\n" + "="*50)
print("Tabla de valores para la función de costo del parque: f(x) = 2x + 8")
# x: número de viajes, f(x): costo total
for x in range(6):
    costo = 2*x + 8
    print(f"Viajes (x): {x} -> Costo f(x): ${costo}")

```

```python
# Definimos dos funciones biyectivas sencillas en conjuntos finitos para ilustrar la inversa y composición
# g: A -> B
A = {1, 2, 3}
B = {'X', 'Y', 'Z'}
g = {1: 'Y', 2: 'Z', 3: 'X'}

# f: B -> C
C = {10, 20, 30}
f = {'X': 30, 'Y': 10, 'Z': 20}

# 1. Función Inversa g^-1: B -> A
g_inversa = {valor: clave for clave, valor in g.items()}
print("Función g original:", g)
print("Función Inversa g^-1:", g_inversa)

# 2. Composición f ∘ g: A -> C
composicion = {}
for a in A:
    # (f ∘ g)(a) = f(g(a))
    valor_intermedio = g[a]
    composicion[a] = f[valor_intermedio]

print("\nComposicion (f o g):", composicion)

```

## 4. Cardinalidad y Numerabilidad

El concepto de **cardinalidad** cuantifica el *"tamaño"* de los conjuntos.

### 4.1 Conjuntos Finitos e Infinitos
* **Conjunto Finito:** Si existe un entero no negativo $n$ tal que el conjunto tiene exactamente $n$ elementos distintos. Su cardinalidad se denota por $|A|$, $\text{card}(A)$ o $\#A$.
* **Conjunto Infinito:** Un conjunto que no es finito (por ejemplo, el conjunto de los números enteros $\mathbb{Z}$).

### 4.2 Comparación de Cardinalidades
Para comparar conjuntos infinitos, Georg Cantor estableció que dos conjuntos $A$ y $B$ tienen la **misma cardinalidad** (se escribe $|A| = |B|$) si y solo si **existe una función biyectiva** de $A$ en $B$.
* $|A| \le |B|$ si existe una función inyectiva de $A$ en $B$.
* $|A| < |B|$ si $|A| \le |B|$ pero no existe ninguna biyección entre ellos (es estrictamente menor).

### 4.3 Conjuntos Numerables (Contables)
Un conjunto $S$ se dice **numerable** o contable si es finito o si tiene la misma cardinalidad que el conjunto de los números enteros positivos $\mathbb{Z}^+$ (o naturales $\mathbb{N}$). La cardinalidad de un conjunto infinito numerable se denota por el símbolo transfinito alef-cero:
$$|S| = \aleph_0$$
Un conjunto que no es numerable se denomina **no numerable** (por ejemplo, los números reales $\mathbb{R}$, cuya cardinalidad es la del continuo: $2^{\aleph_0}$).

#### Ejemplo 1: El conjunto de los enteros impares positivos es numerable.
* **Teorema:** El conjunto $I = \{1, 3, 5, 7, \dots\}$ de los impares positivos tiene cardinalidad $\aleph_0$.
* *Demostración:* Construimos la función $f: \mathbb{Z}^+ \to I$ definida por:
  $$f(n) = 2n - 1$$
  * *Inyectividad:* Si $f(n) = f(m) \implies 2n - 1 = 2m - 1 \implies 2n = 2m \implies n = m$.
  * *Sobreyectividad:* Para cualquier impar positivo $t$, existe un entero $k = (t + 1)/2$. Como $t$ es impar, $t+1$ es par, por lo que $k$ es un entero positivo. Evaluando: $f(k) = 2((t+1)/2) - 1 = t + 1 - 1 = t$.
  * Al ser biyectiva, existe correspondencia perfecta. Así, $|I| = |\mathbb{Z}^+| = \aleph_0$.

### 4.4 Principio de Inclusión-Exclusión
Para calcular la cardinalidad de la unión de conjuntos finitos que pueden no ser disjuntos:
$$|A \cup B| = |A| + |B| - |A \cap B|$$
*(Nota: Si los conjuntos son disjuntos, $A \cap B = \emptyset \implies |A \cap B| = 0$, reduciéndose a la regla de la suma $|A \cup B| = |A| + |B|$).*


```python
import matplotlib.pyplot as plt

# 1. Visualización de la Biyección entre Naturales e Impares (Cardinalidad Infinita)
# Graficamos las flechas de correspondencia para los primeros 8 términos
naturales = list(range(1, 9))
impares = [2*n - 1 for n in naturales]

plt.figure(figsize=(10, 4))
plt.scatter(naturales, [1.5]*len(naturales), color='blue', s=150, zorder=3, label='Naturales ($\mathbb{N}$)')
plt.scatter(impares, [0.5]*len(impares), color='red', s=150, zorder=3, label='Impares ($I$)')

# Dibujamos las flechas de la biyección f(n) = 2n - 1
for n, imp in zip(naturales, impares):
    plt.annotate('', xy=(imp, 0.55), xytext=(n, 1.45),
                 arrowprops=dict(arrowstyle="->", color='gray', lw=1.5, ls='--'))
    plt.text(n, 1.58, f"{n}", ha='center', fontsize=11)
    plt.text(imp, 0.35, f"{imp}", ha='center', fontsize=11)

plt.ylim(0.0, 2.0)
plt.xlim(0, 16)
plt.yticks([0.5, 1.5], ['Impares', 'Naturales'], fontsize=12, fontweight='bold')
plt.title('Biyección de Cantor: $f(n) = 2n - 1$ demostrando $|\mathbb{N}| = |I|$', fontsize=13, fontweight='bold')
plt.grid(False)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# 2. Demostración del Principio de Inclusión-Exclusión
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
A_int_B = A.intersection(B)
A_uni_B = A.union(B)

print("Principio de Inclusion-Exclusion:")
print(f"  - Cardinal de A (|A|): {len(A)}")
print(f"  - Cardinal de B (|B|): {len(B)}")
print(f"  - Cardinal de la interseccion (|A n B|): {len(A_int_B)}")
print(f"  - Cardinal de la union (|A U B|): {len(A_uni_B)}")
# Verificación de la igualdad
print(f"  - ¿|A U B| == |A| + |B| - |A n B|? {len(A_uni_B) == len(A) + len(B) - len(A_int_B)}")

```

## 5. Resumen de la Lección

1. **Conjuntos:** Son la herramienta estructural de las matemáticas para organizar objetos. Pueden definirse por extensión (describiendo sus elementos) o por condición (especificando reglas proposicionales de pertenencia).
2. **Operaciones:** La unión ($A \cup B$), la intersección ($A \cap B$), la diferencia ($A - B$) y la diferencia simétrica ($A \Delta B$) permiten manipular y crear estructuras de datos lógicas.
3. **Funciones:** Definen la correspondencia entre conjuntos. Sus clasificaciones (inyectiva, sobreyectiva y biyectiva) determinan propiedades algebraicas fundamentales como la invertibilidad.
4. **Cardinalidad:** Mide el tamaño de los conjuntos. La existencia de biyecciones es la herramienta clave para comparar infinitos, distinguiendo entre conjuntos infinitos numerables (de cardinal $\aleph_0$) e infinitos no numerables.

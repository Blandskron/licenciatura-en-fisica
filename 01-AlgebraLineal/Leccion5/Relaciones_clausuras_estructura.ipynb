## 1. Relaciones Binarias y sus Propiedades

En matemáticas, una **relación** es un subconjunto de pares ordenados que formaliza la conexión entre elementos de conjuntos.

### 1.1 Definición Formal
Sean $A$ y $B$ dos conjuntos. Una **relación binaria** de $A$ en $B$ es un subconjunto del producto cartesiano $A \times B$.
$$R \subseteq A \times B$$
* Si $(a, b) \in R$, decimos que *"$a$ está relacionado con $b$ por medio de $R$"*, y se escribe $aRb$.
* Si $(a, b) \notin R$, se escribe $a \not R b$.
* **Dominio:** Conjunto de todos los primeros elementos de los pares de $R$:
  $$\text{Dom}(R) = \{a \in A \mid \exists b \in B \ ((a, b) \in R)\}$$
* **Rango:** Conjunto de todos los segundos elementos de los pares de $R$:
  $$\text{Ran}(R) = \{b \in B \mid \exists a \in A \ ((a, b) \in R)\}$$

### 1.2 Operaciones con Relaciones
* **Relación Inversa ($R^{-1}$):** Relación de $B$ en $A$ formada al invertir todos los pares de $R$.
  $$R^{-1} = \{(b, a) \in B \times A \mid (a, b) \in R\}$$
  *(Propiedad: $(R^{-1})^{-1} = R$).*
* **Composición de Relaciones ($S \circ R$):** Sean $R$ una relación de $A$ en $B$ y $S$ una relación de $B$ en $C$. La composición $S \circ R$ es una relación de $A$ en $C$ que consta de los pares $(a, c)$ para los cuales existe al menos un elemento intermedio $b \in B$ tal que $(a, b) \in R$ y $(b, c) \in S$.
  $$S \circ R = \{(a, c) \in A \times C \mid \exists b \in B \ ((a, b) \in R \land (b, c) \in S)\}$$

### 1.3 Propiedades sobre un Único Conjunto $A$
Si la relación está definida sobre un conjunto único (es decir, $R \subseteq A \times A$), estudiamos las siguientes propiedades fundamentales:
1. **Reflexiva:** Cada elemento está relacionado consigo mismo.
   $$\forall a \in A \implies (a, a) \in R$$
2. **Simétrica:** Si un elemento está relacionado con otro, la relación inversa también se cumple.
   $$\forall a, b \in A \ ((a, b) \in R \implies (b, a) \in R)$$
3. **Antisimétrica:** Si dos elementos distintos están relacionados, no pueden estarlo en ambos sentidos a la vez.
   $$\forall a, b \in A \ (((a, b) \in R \land (b, a) \in R) \implies a = b)$$
4. **Transitiva:** Si $a$ está relacionado con $b$ y $b$ con $c$, entonces $a$ está relacionado con $c$.
   $$\forall a, b, c \in A \ (((a, b) \in R \land (b, c) \in R) \implies (a, c) \in R)$$

### 1.4 Relación de Equivalencia
Una relación $R$ sobre $S$ es una **relación de equivalencia** si cumple simultáneamente ser **reflexiva, simétrica y transitiva**. Estas relaciones agrupan a los elementos en particiones disjuntas de "objetos semejantes" llamados clases de equivalencia (por ejemplo, la congruencia modular).

### 1.5 Relaciones n-arias
Generalizan el concepto para $n$ conjuntos. Una relación $n$-aria es un subconjunto de $A_1 \times A_2 \times \dots \times A_n$. El número $n$ es el **grado** de la relación. Son fundamentales en bases de datos relacionales.


```python
# 1. Definición de operaciones de dominio, rango e inversa
def dom_relacion(R):
    return {a for a, b in R}

def ran_relacion(R):
    return {b for a, b in R}

def inversa_relacion(R):
    return {(b, a) for a, b in R}

# 2. Composición de Relaciones S o R
def componer_relaciones(R, S):
    # R es relacion A -> B, S es relacion B -> C
    composicion = set()
    for a, b in R:
        for b2, c in S:
            if b == b2:
                composicion.add((a, c))
    return composicion

# Ejemplo del PDF (Figura 2)
# A = {1, 2, 3, 4}, B = {a, b, c, d}, C = {x, y, z}
R = {(1, 'a'), (2, 'd'), (3, 'a'), (3, 'b'), (3, 'd')}
S = {('b', 'x'), ('b', 'z'), ('c', 'y'), ('d', 'z')}

S_o_R = componer_relaciones(R, S)
print("Relacion R:", R)
print("Relacion S:", S)
print("Composicion S o R (esperado: {(2, 'z'), (3, 'x'), (3, 'z')}):")
print(S_o_R)

```

```python
def verificar_propiedades(A, R):
    # A es el conjunto base, R es un conjunto de tuplas
    
    # 1. Reflexiva: (a,a) en R para todo a en A
    reflexiva = all((a, a) in R for a in A)
    
    # 2. Simétrica: (a,b) en R => (b,a) en R
    simetrica = True
    for a, b in R:
        if (b, a) not in R:
            simetrica = False
            break
            
    # 3. Antisimétrica: (a,b) en R y (b,a) en R => a == b
    antisimetrica = True
    for a, b in R:
        if a != b and (b, a) in R:
            antisimetrica = False
            break
            
    # 4. Transitiva: (a,b) en R y (b,c) en R => (a,c) en R
    transitiva = True
    for a, b in R:
        for b2, c in R:
            if b == b2:
                if (a, c) not in R:
                    transitiva = False
                    break
        if not transitiva:
            break
            
    # Equivalencia
    equivalencia = reflexiva and simetrica and transitiva
    
    return {
        "Reflexiva": reflexiva,
        "Simetrica": simetrica,
        "Antisimetrica": antisimetrica,
        "Transitiva": transitiva,
        "Equivalencia": equivalencia
    }

# Conjunto base A
A_set = {1, 2, 3, 4}

# Relaciones del ejemplo del PDF (Página 5)
R1 = {(1,1),(1,2),(2,1),(2,2),(3,4),(4,1),(4,4)}
R2 = {(1,1),(1,2),(2,1)}
R3 = {(1,1),(1,2),(1,4),(2,1),(2,2),(3,3),(4,1),(4,4)}
R4 = {(2,1),(3,1),(3,2),(4,1),(4,2),(4,3)}
R5 = {(1,1),(1,2),(1,3),(1,4),(2,2),(2,3),(2,4),(3,3),(3,4),(4,4)}
R6 = {(3,4)}

relaciones = [R1, R2, R3, R4, R5, R6]
for idx, R_i in enumerate(relaciones, 1):
    props = verificar_propiedades(A_set, R_i)
    print(f"Propiedades de R{idx}:")
    for prop, val in props.items():
        print(f"  - {prop}: {val}")
    print("-" * 30)

```

## 2. Representación de Relaciones (Matrices y Dígrafos)

Las relaciones sobre conjuntos finitos pueden representarse de forma compacta y operar mediante estructuras computacionales y gráficas.

### 2.1 Representación Matricial (Matrices de la Relación)
Sean $A = \{a_1, \dots, a_m\}$ y $B = \{b_1, \dots, b_n\}$ dos conjuntos ordenados. Representamos una relación $R \subseteq A \times B$ mediante una matriz booleana $M_R = [m_{ij}]$ de tamaño $m \times n$, donde:
$$m_{ij} = \begin{cases} 1 & \text{si } (a_i, b_j) \in R \\ 0 & \text{si } (a_i, b_j) \notin R \end{cases}$$

#### Propiedades en la Matriz ($M_R$ sobre $A \times A$)
* **Reflexiva:** Todos los elementos de la diagonal principal son 1 ($m_{ii} = 1$).
* **Simétrica:** La matriz es simétrica ($M_R = M_R^T$, es decir, $m_{ij} = m_{ji}$).
* **Antisimétrica:** Si $m_{ij} = 1$ para $i \neq j$, entonces $m_{ji} = 0$.

### 2.2 Representación mediante Grafos Dirigidos (Dígrafos)
Un **grafo dirigido** (o dígrafo) consta de un conjunto $V$ de vértices (nodos) y un conjunto $E$ de aristas orientadas (arcos) formadas por pares ordenados de vértices.
* Si $(a, b) \in R$, dibujamos una flecha (arista orientada) del vértice $a$ al vértice $b$.
* Si $(a, a) \in R$, dibujamos un **bucle** en el vértice $a$.

#### Interpretación de Propiedades en el Dígrafo
* **Reflexiva:** Hay un bucle en cada vértice del dígrafo.
* **Simétrica:** Para toda arista que conecta dos nodos distintos en un sentido, existe otra en sentido opuesto.
* **Antisimétrica:** No hay dos nodos distintos conectados por aristas en ambos sentidos.
* **Transitiva:** Si hay una arista de $x$ a $y$ y otra de $y$ a $z$, necesariamente existe una arista directa de $x$ a $z$ (cierre de triángulos orientados).


```python
import numpy as np

def relacion_a_matriz(A, B, R):
    list_A = sorted(list(A))
    list_B = sorted(list(B))
    m, n = len(list_A), len(list_B)
    M = np.zeros((m, n), dtype=int)
    for i, a in enumerate(list_A):
        for j, b in enumerate(list_B):
            if (a, b) in R:
                M[i, j] = 1
    return M

def matriz_a_relacion(A, B, M):
    list_A = sorted(list(A))
    list_B = sorted(list(B))
    R = set()
    for i in range(len(list_A)):
        for j in range(len(list_B)):
            if M[i, j] == 1:
                R.add((list_A[i], list_B[j]))
    return R

# Ejemplo del PDF (Página 6)
# A = {1, 2, 3}, B = {'x', 'y', 'z'}
# R = {(1,'y'), (1,'z'), (3,'z')}
A_s = {1, 2, 3}
B_s = {'x', 'y', 'z'}
R_s = {(1, 'y'), (1, 'z'), (3, 'z')}

M_R = relacion_a_matriz(A_s, B_s, R_s)
print("Matriz de la relacion R:")
print(M_R)

# Reconstruir la relación a partir de la matriz
R_reconstruido = matriz_a_relacion(A_s, B_s, M_R)
print("\nRelacion reconstruida:", R_reconstruido)
print("¿Coinciden?", R_reconstruido == R_s)

```

## 3. Cierre (Clausura) de una Relación

El **cierre** de una relación $R$ respecto a una propiedad $P$ es la relación **más pequeña** que contiene a $R$ y que cumple con la propiedad $P$ (si es que existe).

### 3.1 Tipos de Cierres

1. **Cierre Reflexivo:** Se obtiene agregando a $R$ todos los pares del tipo $(a, a)$ para todo $a \in A$. Se denota por $R \cup \Delta$, donde $\Delta$ es la diagonal:
   $$\Delta = \{(a, a) \mid a \in A\}$$
2. **Cierre Simétrico:** Se obtiene agregando a $R$ todos los pares inversos $(b, a)$ de los pares $(a, b) \in R$. Equivale a la unión de la relación con su inversa:
   $$\text{Cierre Simétrico} = R \cup R^{-1}$$
3. **Cierre Transitivo ($R^*$):**
   * **Teorema de Conectividad:** Si un conjunto $A$ tiene cardinal $n$, el cierre transitivo $R^*$ (que contiene a todos los pares conectados por caminos de cualquier longitud) equivale a la unión de las primeras $n$ potencias de la relación:
     $$R^* = \bigcup_{i=1}^n R^i = R \cup R^2 \cup \dots \cup R^n$$
   * **Caminos con Matrices Booleanas:** Si $M_R$ es la matriz booleana de la relación $R$, la matriz del cierre transitivo $M_{R^*}$ se obtiene calculando la disyunción lógica (OR elemento a elemento) de sus potencias booleanas:
     $$M_{R^*} = M_R \lor M_R^{[2]} \lor \dots \lor M_R^{[n]}$$

### 3.2 Algoritmo de Warshall
El cálculo de potencias booleanas sucesivas para obtener el cierre transitivo requiere muchas operaciones. Stephen Warshall desarrolló un algoritmo altamente eficiente con complejidad temporal de $O(n^3)$ que trabaja directamente sobre la matriz booleana:

Sea $W$ la matriz inicial igual a $M_R$. Para cada nodo intermedio $k$ desde $0$ hasta $n-1$:
$$w_{ij} = w_{ij} \lor (w_{ik} \land w_{kj})$$
Al finalizar, la matriz $W$ contiene el cierre transitivo de la relación.


```python
# 1. Cierre Reflexivo
def cierre_reflexivo(A, R):
    cierre = set(R)
    for a in A:
        cierre.add((a, a))
    return cierre

# 2. Cierre Simétrico
def cierre_simetrico(R):
    cierre = set(R)
    inversa = inversa_relacion(R)
    return closure_union(cierre, inversa)

def closure_union(s1, s2):
    return s1.union(s2)

# Ejemplo del PDF (Página 8)
# A = {1, 2, 3}, R = {(1,1),(1,2),(2,1),(3,2)}
A_cierre = {1, 2, 3}
R_cierre = {(1, 1), (1, 2), (2, 1), (3, 2)}

c_reflexivo = cierre_reflexivo(A_cierre, R_cierre)
print("Cierre Reflexivo (esperado: {(1,1),(1,2),(2,1),(3,2),(2,2),(3,3)}):")
print(c_reflexivo)
print("-" * 50)

# Ejemplo del PDF (Página 9)
# R = {(1,1), (1,2), (2,3), (3,1), (3,2)}
R_sim = {(1, 1), (1, 2), (2, 3), (3, 1), (3, 2)}
c_simetrico = cierre_simetrico(R_sim)
print("Cierre Simetrico:")
print(c_simetrico)

```

```python
# 1. Cierre Transitivo por Potencias Booleanas
def producto_booleano(A, B):
    m, k = A.shape
    k2, n = B.shape
    C = np.zeros((m, n), dtype=int)
    for i in range(m):
        for j in range(n):
            val_or = 0
            for x in range(k):
                val_or = val_or | (A[i, x] & B[x, j])
            C[i, j] = val_or
    return C

def cierre_transitivo_potencias(M):
    n = M.shape[0]
    res = M.copy()
    temp = M.copy()
    for _ in range(2, n + 1):
        temp = producto_booleano(temp, M)
        res = res | temp
    return res

# 2. Cierre Transitivo mediante el Algoritmo de Warshall
def algoritmo_warshall(M):
    W = M.copy()
    n = W.shape[0]
    for k in range(n):
        for i in range(n):
            for j in range(n):
                # Si hay camino de i a k y de k a j, entonces hay camino de i a j
                W[i, j] = W[i, j] | (W[i, k] & W[k, j])
    return W

# Ejemplo del PDF (Página 10)
# M_R = [[1, 0, 1], [0, 1, 0], [1, 1, 0]]
M_R_test = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 0]
], dtype=int)

print("Matriz de relacion original M_R:")
print(M_R_test)
print("-" * 50)

# Calculamos el cierre transitivo usando potencias
ct_pot = cierre_transitivo_potencias(M_R_test)
print("Cierre Transitivo por Potencias Booleanas:")
print(ct_pot)
print("-" * 50)

# Calculamos el cierre transitivo usando Warshall
ct_warshall = algoritmo_warshall(M_R_test)
print("Cierre Transitivo usando el Algoritmo de Warshall:")
print(ct_warshall)
print("-" * 50)

# Comparamos ambos métodos
print("¿Ambos metodos coinciden?", np.array_equal(ct_pot, ct_warshall))

```

## 4. Resumen de la Lección

1. **Relación Binaria:** Subconjunto de $A \times B$ que formaliza enlaces entre elementos. Tiene dominio, rango e inversa.
2. **Propiedades:** Reflexiva (conectado consigo mismo), simétrica (enlace de doble vía), antisimétrica (no hay enlaces recíprocos entre distintos) y transitiva (conectividad indirecta).
3. **Equivalencia:** Cumple reflexividad, simetría y transitividad, y divide un conjunto en clases semejantes.
4. **Representación:** Las matrices booleanas y los grafos dirigidos (dígrafos) permiten visualizar y codificar relaciones de forma óptima.
5. **Clausuras (Cierres):** Extienden una relación para forzar el cumplimiento de una propiedad. El cierre transitivo calcula la conectividad total y se obtiene eficientemente mediante el algoritmo de Warshall.

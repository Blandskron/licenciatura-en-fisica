## 1. Fundamentos Matemáticos: Geometría Analítica y Álgebra Elemental

El Álgebra Lineal extiende las ecuaciones algebraicas simples al espacio multidimensional. 

### 1.1 Ecuación de la Recta
Una recta en el plano bidimensional se define mediante la ecuación lineal de primer orden:
$$y = m \cdot x + b$$
* Donde $m$ representa la **pendiente** (tasa de cambio de $y$ con respecto a $x$).
* $b$ representa la **intersección con el eje Y** (el valor de $y$ cuando $x=0$).

### 1.2 Sistemas de Coordenadas
* **Plano 2D:** Cada punto se localiza mediante un par ordenado $(x, y)$ respecto a un origen $(0,0)$.
* **Espacio 3D:** Cada punto se localiza mediante una terna ordenada $(x, y, z)$ respecto a un origen $(0,0,0)$.
* **Vectores Geométricos:** Un vector es una entidad que posee **magnitud** (longitud del segmento), **dirección** (recta sobre la que se apoya) y **sentido** (indicado por la punta de la flecha). Se representa típicamente desde el origen hacia las coordenadas del punto de destino.
* **Distancia Euclidiana:** La distancia entre dos puntos $P_1(x_1, y_1)$ y $P_2(x_2, y_2)$ en el plano es:
  $$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Generar rango de valores para x
x_recta = np.linspace(-5, 5, 100)
# Definir una pendiente m y una intersección b
m = 1.5
b = -2.0
# Ecuación de la recta
y_recta = m * x_recta + b

plt.figure(figsize=(8, 5))
plt.plot(x_recta, y_recta, 'b-', label=f'y = {m}x + {b}')
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.axvline(0, color='black', linewidth=1, linestyle='--')

# Graficar un vector desde el origen (0,0) hasta el punto (2, 1)
plt.quiver(0, 0, 2, 1, angles='xy', scale_units='xy', scale=1, color='red', label='Vector (2, 1)')
plt.scatter(2, 1, color='red', zorder=5)

plt.title("Geometría Analítica: Rectas y Vectores en el Plano 2D", fontsize=11, fontweight='bold')
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()

```

## 2. Fundamentos de Lógica y Álgebra Booleana

El álgebra lineal y computacional requiere el dominio de la lógica binaria para estructurar condicionales y operaciones lógicas:

### 2.1 Conectivos Lógicos Fundamentales
Dadas dos proposiciones $A$ y $B$:
* **Conjunción (AND - $\land$):** Es verdadera si y solo si ambas proposiciones son verdaderas. En Python se escribe como `and`.
* **Disyunción (OR - $\lor$):** Es verdadera si al menos una de las proposiciones es verdadera. En Python se escribe como `or`.
* **Negación (NOT - $\neg$):** Invierte el valor de verdad. En Python se escribe como `not`.

### 2.2 Tablas de Verdad
Las tablas de verdad listan todos los posibles escenarios de valores de verdad para evaluar el resultado de una operación lógica.


```python
# Construcción programática de tablas de verdad en Python
valores = [True, False]

print("-" * 40)
print(f"{'A':<8} | {'B':<8} | {'A and B':<10} | {'A or B':<10}")
print("-" * 40)
for A in valores:
    for B in valores:
        res_and = A and B
        res_or = A or B
        print(f"{str(A):<8} | {str(B):<8} | {str(res_and):<10} | {str(res_or):<10}")
print("-" * 40)

```

## 3. Introducción Básica a la Programación en Python

Python es el lenguaje estándar de la física matemática y la computación científica gracias a su legibilidad y robusto ecosistema de librerías. Para este módulo debes dominar:
* **Declaración de Variables:** Asignación dinámica de tipos (`x = 5`, `nombre = "Física"`).
* **Colecciones (Listas y Diccionarios):** `lista = [1, 2, 3]`, acceder por índice `lista[0]`.
* **Estructuras de Control de Flujo:**
  - Condicionales (`if`, `elif`, `else`).
  - Bucles de repetición (`for` para iterar rangos o colecciones, `while` para condiciones dinámicas).
* **Funciones (`def`):** Modularización de código con parámetros de entrada y retorno de resultados (`return`).


```python
# Ejemplo de flujo de control y funciones en Python
def evaluar_temperatura(temp_celsius):
    """
    Clasifica la temperatura y retorna un mensaje
    """
    if temp_celsius < 10.0:
        return "Frío"
    elif temp_celsius <= 25.0:
        return "Templado"
    else:
        return "Cálido"

# Lista de temperaturas simuladas
temperaturas = [5.5, 18.2, 30.0, 12.0]

print("--- Clasificación de Temperaturas ---")
for temp in temperaturas:
    estado = evaluar_temperatura(temp)
    print(f"Temperatura: {temp:>5.1f} °C -> Estado: {estado}")

```

## 4. Computación Científica con NumPy

Las listas de Python son lentas e ineficientes para cálculos numéricos masivos porque no se almacenan de forma contigua en memoria. Para resolver esto, la librería **NumPy** introduce los arreglos vectorizados (`ndarrays`):
* **Vector (1D Array):** Arreglo de una sola dimensión.
* **Matriz (2D Array):** Arreglo bidimensional estructurado en filas y columnas.
* **Vectorización:** Las operaciones elementales se aplican a todo el arreglo de forma simultánea a velocidad de código nativo (C/C++), evitando bucles lentos en Python.


```python
import numpy as np

# 1. Crear vectores en 1D
vec1 = np.array([1, 2, 3])
vec2 = np.array([10, 20, 30])

# Operación element-wise (elemento a elemento)
suma_vec = vec1 + vec2
producto_escalar = np.dot(vec1, vec2) # 1*10 + 2*20 + 3*30 = 140

print("--- Computación Científica con NumPy ---")
print("Vector 1:", vec1)
print("Vector 2:", vec2)
print("Suma elemento a elemento:", suma_vec)
print("Producto Escalar (dot product):", producto_escalar)

# 2. Crear una matriz en 2D (Fila, Columna)
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print("\nMatriz 2D:\n", matriz)
print("Forma (Shape) de la matriz (filas, columnas):", matriz.shape)
print("Suma total de elementos de la matriz:", np.sum(matriz))

```

## 5. Graficación Científica con Matplotlib

La visualización de datos y funciones es crítica en física matemática para contrastar teorías con resultados numéricos. La librería estándar de graficación es **Matplotlib**:
* Para graficar una función matemática continua $f(x)$, generamos una grilla densa de puntos espaciados en $x$ usando `np.linspace`, calculamos sus valores $y = f(x)$, y los unimos en un trazo continuo usando `plt.plot`.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 200 puntos entre 0 y 2*pi
x_rad = np.linspace(0, 2 * np.pi, 200)
y_seno = np.sin(x_rad)
y_coseno = np.cos(x_rad)

plt.figure(figsize=(10, 5))

# Trazar las curvas
plt.plot(x_rad, y_seno, 'b-', linewidth=2, label='f(x) = Seno(x)')
plt.plot(x_rad, y_coseno, 'r--', linewidth=2, label='f(x) = Coseno(x)')

# Personalizar el gráfico
plt.title("Visualización de Funciones Trigonométricas continuas", fontsize=12, fontweight='bold')
plt.xlabel("Eje X (Radianes)", fontsize=10)
plt.ylabel("Amplitud", fontsize=10)
plt.axhline(0, color='black', linewidth=0.8) # Línea base en y=0
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper right')

plt.show()

```

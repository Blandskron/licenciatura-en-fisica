## 1. Álgebra Elemental: Exponentes, Factorización y Racionalización

El álgebra es el motor operativo del cálculo. La mayor parte de los errores al resolver límites, derivadas o integrales no se deben a fallas en los conceptos del cálculo, sino a imprecisiones en la manipulación algebraica básica.

### 1.1 Leyes de los Exponentes y Radicales
Para cualquier base real $a, b \neq 0$ y exponentes reales $n, m$, se cumplen las siguientes leyes fundamentales:

| Regla | Expresión Matemática | Descripción |
| :--- | :--- | :--- |
| **Producto de Potencias** | $a^n \cdot a^m = a^{n+m}$ | Al multiplicar potencias de igual base, se suman los exponentes. |
| **Cociente de Potencias** | $\frac{a^n}{a^m} = a^{n-m}$ | Al dividir potencias de igual base, se restan los exponentes. |
| **Potencia de una Potencia** | $(a^n)^m = a^{n \cdot m}$ | Se multiplican los exponentes. |
| **Potencia de un Producto** | $(a \cdot b)^n = a^n \cdot b^n$ | El exponente se distribuye al producto. |
| **Exponente Negativo** | $a^{-n} = \frac{1}{a^n}$ | La potencia con exponente negativo es la inversa con exponente positivo. |
| **Exponente Fraccionario (Radical)** | $a^{n/m} = \sqrt[m]{a^n}$ | El denominador del exponente indica el índice del radical. |

### 1.2 Técnicas Clave de Factorización
Factorizar consiste en expresar una suma o resta como un producto de factores. Las fórmulas algebraicas esenciales son:
1. **Diferencia de Cuadrados**:
   $$a^2 - b^2 = (a - b)(a + b)$$
2. **Trinomio Cuadrado Perfecto**:
   $$a^2 \pm 2ab + b^2 = (a \pm b)^2$$
3. **Diferencia y Suma de Cubos**:
   $$a^3 - b^3 = (a - b)(a^2 + ab + b^2)$$
   $$a^3 + b^3 = (a + b)(a^2 - ab + b^2)$$

### 1.3 Racionalización
Racionalizar consiste en eliminar los radicales del denominador (o del numerador en el caso de cálculo de ciertos límites indeterminados). Para ello, multiplicamos arriba y abajo por un factor adecuado (usualmente el conjugado):
- **Caso de raíz única**: Para eliminar $\sqrt{a}$, multiplicamos por $\frac{\sqrt{a}}{\sqrt{a}}$.
- **Caso binomial (Conjugado)**: Para eliminar $(\sqrt{a} - \sqrt{b})$, multiplicamos por su conjugado $(\sqrt{a} + \sqrt{b})$:
  $$\frac{c}{\sqrt{a} - \sqrt{b}} = \frac{c(\sqrt{a} + \sqrt{b})}{(\sqrt{a} - \sqrt{b})(\sqrt{a} + \sqrt{b})} = \frac{c(\sqrt{a} + \sqrt{b})}{a - b}$$


## 2. Trigonometría Fundamental y el Círculo Unitario

Las funciones trigonométricas son fundamentales en física para modelar fenómenos periódicos (movimiento armónico simple, ondas, corriente alterna, óptica).

### 2.1 Medida de Ángulos: Grados y Radianes
Un **radián** es la medida de un ángulo central subtendido por un arco cuya longitud es igual al radio del círculo. Dado que una circunferencia completa tiene longitud $2\pi r$, se tiene que:

$$360^{\circ} = 2\pi \text{ rad} \implies 1 \text{ rad} = \frac{180^{\circ}}{\pi} \approx 57.296^{\circ}$$

Para convertir de grados a radianes, multiplicamos por $\frac{\pi}{180^{\circ}}$. Para convertir de radianes a grados, multiplicamos por $\frac{180^{\circ}}{\pi}$.

### 2.2 Definiciones sobre el Círculo Unitario
El círculo unitario es aquel con radio $r = 1$ centrado en el origen $(0,0)$. Para cualquier ángulo $\theta$ en posición estándar, el rayo terminal interseca al círculo en el punto $(x, y)$. Las funciones trigonométricas se definen como:

$$\cos(\theta) = x, \quad \sin(\theta) = y, \quad \tan(\theta) = \frac{y}{x} \quad (x \neq 0)$$

Esto da origen a la **Identidad Pitagórica Fundamental** (aplicando el teorema de Pitágoras en el círculo unitario):

$$\sin^2(\theta) + \cos^2(\theta) = 1$$

### 2.3 Identidades Trigonométricas Clave
Además de la identidad fundamental, el estudiante debe dominar las siguientes relaciones:
- **Fórmulas de Suma de Ángulos**:
  $$\sin(a \pm b) = \sin(a)\cos(b) \pm \cos(a)\sin(b)$$
  $$\cos(a \pm b) = \cos(a)\cos(b) \mp \sin(a)\sin(b)$$
- **Fórmulas de Ángulo Doble**:
  $$\sin(2\theta) = 2\sin(\theta)\cos(\theta)$$
  $$\cos(2\theta) = \cos^2(\theta) - \sin^2(\theta) = 2\cos^2(\theta) - 1 = 1 - 2\sin^2(\theta)$$
- **Fórmulas de Ángulo Medio (Reducción de Potencia)**:
  $$\sin^2(\theta) = \frac{1 - \cos(2\theta)}{2}, \quad \cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}$$
  *(Estas identidades son cruciales en el cálculo de integrales trigonométricas, Lección 5)*.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.titlesize': 14
})

# Definición de ángulo de ejemplo: theta = 60 grados (pi/3 radianes)
theta_val = np.pi / 3.0
x_pt, y_pt = np.cos(theta_val), np.sin(theta_val)

# Creación de la Visualización 1: Círculo unitario y ondas trigonométricas
fig, axes = plt.subplots(1, 2, figsize=(14, 6.5))

# Panel Izquierdo: Círculo Unitario
phi = np.linspace(0, 2*np.pi, 200)
axes[0].plot(np.cos(phi), np.sin(phi), color='black', linewidth=1.5, label='Círculo Unitario ($r=1$)')
axes[0].axhline(0, color='gray', linestyle='-', linewidth=0.8)
axes[0].axvline(0, color='gray', linestyle='-', linewidth=0.8)

axes[0].plot([0, x_pt], [0, y_pt], color='purple', linewidth=2.5, marker='o', label='Rayo terminal')
axes[0].plot([x_pt, x_pt], [0, y_pt], color='#d62728', linewidth=2.0, linestyle='--', label=f'$\\\\sin(\\\\theta) = y \\\\approx {y_pt:.4f}$')
axes[0].plot([0, x_pt], [0, 0], color='#1f77b4', linewidth=2.5, label=f'$\\\\cos(\\\\theta) = x \\\\approx {x_pt:.4f}$')

# Sombreado de ángulo
ang_vals = np.linspace(0, theta_val, 50)
axes[0].fill_between(0.2*np.cos(ang_vals), 0.2*np.sin(ang_vals), color='purple', alpha=0.2, label=f'$\\\\theta = \\\\pi/3$ ($60^\\\\circ$)')

axes[0].set_title('Representación en el Círculo Unitario', pad=12)
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
axes[0].set_xlim(-1.2, 1.2)
axes[0].set_ylim(-1.2, 1.2)
axes[0].set_aspect('equal')
axes[0].legend(loc='upper left', frameon=True, facecolor='white', edgecolor='none')
axes[0].grid(True, linestyle=':', alpha=0.5)

# Panel Derecho: Trazado de las Ondas
theta_domain = np.linspace(0, 2*np.pi, 500)
axes[1].plot(theta_domain, np.sin(theta_domain), color='#d62728', linewidth=2.0, label='$\\\\sin(\\\\theta)$')
axes[1].plot(theta_domain, np.cos(theta_domain), color='#1f77b4', linewidth=2.0, label='$\\\\cos(\\\\theta)$')
axes[1].axhline(0, color='gray', linestyle='-', linewidth=0.5)

# Marcación del punto del ángulo actual
axes[1].scatter(theta_val, y_pt, color='#d62728', s=80, zorder=5)
axes[1].scatter(theta_val, x_pt, color='#1f77b4', s=80, zorder=5)
axes[1].axvline(theta_val, color='purple', linestyle=':', linewidth=1.5)

axes[1].set_title('Gráficas de las Funciones Trigonométricas', pad=12)
axes[1].set_xlabel('Ángulo $\theta$ (Radianes)')
axes[1].set_ylabel('Valor')
axes[1].set_xlim(0, 2*np.pi)
axes[1].set_ylim(-1.1, 1.1)
axes[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axes[1].set_xticklabels(['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$'])
axes[1].legend(loc='lower left', frameon=True, facecolor='white', edgecolor='none')
axes[1].grid(True, linestyle=':', alpha=0.5)

plt.suptitle('Visualización 1: El Círculo Unitario y la Generación de Ondas', y=0.98)
plt.tight_layout()
plt.show()
```

## 3. Geometría Analítica en el Plano Cartesiano

La geometría analítica conecta la geometría clásica con el álgebra numérica, permitiendo describir curvas geométricas mediante ecuaciones algebraicas. Es la cuna de los conceptos de derivada (pendiente) e integración definida (recintos planos).

### 3.1 Distancia entre Puntos y Punto Medio
Para dos puntos $P_1(x_1, y_1)$ y $P_2(x_2, y_2)$ en el plano cartesiano $\mathbb{R}^2$:
- **Distancia euclidiana ($d$)**: Proviene directamente del teorema de Pitágoras:
  $$d(P_1, P_2) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$
- **Punto medio ($M$)**: Coordenadas promedio de los extremos:
  $$M = \left( \frac{x_1 + x_2}{2}, \frac{y_1 + y_2}{2} \right)$$

### 3.2 La Pendiente ($m$) de una Recta y su Significado
La **pendiente** de una recta no vertical mide su inclinación o grado de inclinación. Se define formalmente como la razón del cambio en la coordenada vertical ($y$) al cambio en la coordenada horizontal ($x$):

$$m = \frac{\text{desplazamiento vertical}}{\text{desplazamiento horizontal}} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} \quad (x_1 \neq x_2)$$

> **Interpretación Física Crucial**: La pendiente representa la **tasa de cambio promedio** de la variable $y$ respecto a $x$. Si la recta es una curva de movimiento donde el eje X es el tiempo y el eje Y es la posición, la pendiente representa la **velocidad constante**. En cálculo, generalizaremos esta tasa de cambio constante (lineal) a tasas de cambio variables instantáneas (rectas tangentes a curvas) mediante el concepto de derivada (Lección 2).

### 3.3 Ecuaciones de la Recta
Dependiendo de los datos conocidos, representamos la ecuación lineal de una recta en las siguientes formas:
- **Forma Punto-Pendiente**: Útil cuando se conoce un punto $P_1(x_1, y_1)$ y la pendiente $m$:
  $$y - y_1 = m(x - x_1)$$
- **Forma Pendiente-Intercepto**: Útil cuando se conoce la pendiente $m$ y el intercepto con el eje Y $(0, b)$:
  $$y = mx + b$$
- **Ecuación General**: Forma lineal homogénea:
  $$Ax + By + C = 0$$

```python
# Creación de la Visualización 2: Pendiente y tasa de cambio
x_vals = np.linspace(-2.0, 5.0, 200)

plt.figure(figsize=(11, 6.5))

# Trazado de rectas con diferentes pendientes
plt.plot(x_vals, 2.0*x_vals - 1.0, label='Pendiente $m = 2.0$ (Crecimiento rápido)', color='#1f77b4', linewidth=2.5)
plt.plot(x_vals, 0.5*x_vals + 1.0, label='Pendiente $m = 0.5$ (Crecimiento lento)', color='#2ca02c', linewidth=2.0)
plt.plot(x_vals, 0.0*x_vals + 2.0, label='Pendiente $m = 0.0$ (Constante)', color='#9467bd', linewidth=1.8, linestyle='--')
plt.plot(x_vals, -1.0*x_vals + 3.0, label='Pendiente $m = -1.0$ (Decrecimiento)', color='#d62728', linewidth=2.0)

# Ilustrar el triángulo de pendiente (rise over run) para m = 2
# x_1 = 1, y_1 = 1; x_2 = 3, y_2 = 5
plt.plot([1.0, 3.0], [1.0, 1.0], color='black', linestyle=':', linewidth=1.5)
plt.plot([3.0, 3.0], [1.0, 5.0], color='black', linestyle=':', linewidth=1.5)
plt.scatter([1.0, 3.0], [1.0, 5.0], color='black', s=60, zorder=5)

plt.text(2.0, 0.5, '$\Delta x = 2.0$', fontsize=11, ha='center')
plt.text(3.3, 3.0, '$\Delta y = 4.0$', fontsize=11, va='center')
plt.text(1.8, 3.2, '$m = \\frac{\\Delta y}{\\Delta x} = 2.0$', fontsize=12, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

plt.title('Interpretación Geométrica de la Pendiente como Tasa de Cambio', pad=12)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.5, 4.5)
plt.ylim(-2.0, 6.0)
plt.axhline(0, color='gray', linestyle='-', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='-', linewidth=0.5)
plt.legend(frameon=True, facecolor='white', edgecolor='none')
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

## 4. Concepto de Función y Catálogo de Funciones Elementales

El cálculo infinitesimal es, en esencia, la matemática de las **funciones** (Lección 3).

### 4.1 Definición de Función, Dominio y Rango
- **Función ($f$)**: Es una regla de correspondencia que asigna a cada elemento $x$ de un conjunto $D$ (denominado **Dominio**) exactamente un único elemento $y = f(x)$ de un conjunto $E$ (denominado **Rango** o Codominio).
- **Dominio**: El conjunto de todos los valores de entrada permitidos para los cuales la función está definida. En funciones reales, las restricciones típicas son:
  - Evitar divisiones entre cero (denominador $\neq 0$).
  - Evitar raíces pares de números negativos (radicando $\ge 0$).
  - Evitar logaritmos de números menores o iguales a cero (argumento $> 0$).
- **Rango**: El conjunto de todos los valores resultantes de la salida de la función.

### 4.2 Catálogo de Funciones de Referencia
Un físico matemático debe reconocer la morfología gráfica e identificar el dominio y rango de las siguientes familias:

1. **Funciones Cuadráticas ($f(x) = ax^2 + bx + c$)**:
   - Representan parábolas. El dominio es siempre $\mathbb{R}$. Tienen un punto crítico extremo (máximo o mínimo global) en su vértice.
2. **Funciones Racionales Básicas ($f(x) = 1/x$)**:
   - Poseen una discontinuidad esencial en $x = 0$, que se traduce geométricamente en una asíntota vertical y una asíntota horizontal en $y = 0$.
3. **Funciones Exponenciales ($f(x) = e^x$) y Logarítmicas ($f(x) = \ln(x)$)**:
   - Son funciones inversas entre sí. Su simetría reflexiva respecto a la recta $y = x$ es perfecta.
   - El dominio de $e^x$ es $\mathbb{R}$ y su rango es $(0, \infty)$.
   - El dominio de $\ln(x)$ es $(0, \infty)$ y su rango es $\mathbb{R}$.


```python
# Creación de la Visualización 3: Catálogo de funciones elementales
fig, axes = plt.subplots(2, 2, figsize=(13, 11))

# Subplot 1: Función Cuadrática y = x^2
x_quad = np.linspace(-3, 3, 200)
axes[0, 0].plot(x_quad, x_quad**2, color='#1f77b4', linewidth=2.0, label='$y=x^2$')
axes[0, 0].scatter(0, 0, color='orange', s=70, zorder=5, label='Vértice (Mínimo)')
axes[0, 0].set_title('Función Cuadrática (Parábola)', pad=8)
axes[0, 0].set_xlim(-3.1, 3.1)
axes[0, 0].set_ylim(-0.5, 9.0)
axes[0, 0].legend(frameon=True, facecolor='white', edgecolor='none')
axes[0, 0].grid(True, linestyle=':', alpha=0.5)

# Subplot 2: Función Racional y = 1/x
# Dividimos en dos partes para evitar dibujar la línea vertical de salto en x=0
x_rac1 = np.linspace(-3.0, -0.1, 150)
x_rac2 = np.linspace(0.1, 3.0, 150)
axes[0, 1].plot(x_rac1, 1.0/x_rac1, color='#2ca02c', linewidth=2.0)
axes[0, 1].plot(x_rac2, 1.0/x_rac2, color='#2ca02c', linewidth=2.0, label='$y=1/x$')
axes[0, 1].axvline(0, color='red', linestyle=':', label='Asíntota Vertical $x=0$')
axes[0, 1].axhline(0, color='blue', linestyle=':', label='Asíntota Horizontal $y=0$')
axes[0, 1].set_title('Función Racional (Hipérbola)', pad=8)
axes[0, 1].set_xlim(-3.1, 3.1)
axes[0, 1].set_ylim(-10, 10)
axes[0, 1].legend(frameon=True, facecolor='white', edgecolor='none')
axes[0, 1].grid(True, linestyle=':', alpha=0.5)

x_exp = np.linspace(-3.0, 2.2, 200)
x_log = np.linspace(0.05, 9.0, 200)
axes[1, 0].plot(x_exp, np.exp(x_exp), color='#9467bd', linewidth=2.0, label='$y=e^x$')
axes[1, 0].plot(x_log, np.log(x_log), color='#ff7f0e', linewidth=2.0, label='$y=\\\\ln(x)$')
axes[1, 0].plot([-3, 9], [-3, 9], color='gray', linestyle=':', label='Línea de simetría $y=x$')
axes[1, 0].set_title('Exponencial y Logaritmo (Funciones Inversas)', pad=8)
axes[1, 0].set_xlim(-3.0, 6.0)
axes[1, 0].set_ylim(-3.0, 6.0)
axes[1, 0].set_aspect('equal')
axes[1, 0].legend(frameon=True, facecolor='white', edgecolor='none')
axes[1, 0].grid(True, linestyle=':', alpha=0.5)

# Subplot 4: Función Tangente y = tan(x)
x_tan = np.linspace(-1.5, 1.5, 300) # Evitamos pi/2 approx 1.57
axes[1, 1].plot(x_tan, np.tan(x_tan), color='#d62728', linewidth=2.0, label='$y=\\\\tan(x)$')
axes[1, 1].axvline(-np.pi/2, color='red', linestyle=':', label='Asíntota $x=-\\\\pi/2$')
axes[1, 1].axvline(np.pi/2, color='red', linestyle=':', label='Asíntota $x=\\\\pi/2$')
axes[1, 1].set_title('Función Trigonométrica Tangente', pad=8)
axes[1, 1].set_xlim(-2.0, 2.0)
axes[1, 1].set_ylim(-6.0, 6.0)
axes[1, 1].legend(frameon=True, facecolor='white', edgecolor='none')
axes[1, 1].grid(True, linestyle=':', alpha=0.5)

plt.suptitle('Visualización 3: Catálogo Morfológico de las Funciones Reales Elementales', y=0.98)
plt.tight_layout()
plt.show()
```

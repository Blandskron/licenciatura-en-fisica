## 1. Representación de la Información y Aproximaciones

### 1.1 Representación de la Información
En cualquier dispositivo computacional, la memoria es **finita**. Esto condiciona la representación numérica, forzando aproximaciones para números irracionales (como $\pi$, $e$, $\sqrt{2}$) o racionales con parte fraccionaria infinita. Un **formato de representación** establece el número de dígitos enteros y fraccionarios disponibles.

### 1.2 Rango y Precisión
- **Rango**: Es el intervalo cerrado $[a, b]$ que delimita los valores representables más pequeños y más grandes. Si un formato en base 10 tiene la forma $x_1 x_0, x_{-1} x_{-2}$ (2 enteros y 2 fraccionarios), su rango es $[00.00, 99.99]$. Los números fuera del rango no son representables.
- **Precisión ($P$)**: Es la distancia entre dos números representables consecutivos. En el formato anterior, la precisión es $P = 0.01$. Un formato permite distinguir $1.52$ de $1.53$, pero no de $1.5234$.
- **Error de Representación**: La distancia en valor absoluto entre el número real $x$ y su aproximación representable $\hat{x}$:
  $$\epsilon = |x - \hat{x}|$$

### 1.3 Criterios de Aproximación: Truncamiento y Redondeo
1. **Truncamiento (o Chopping)**: Consiste en ignorar o despreciar todos los dígitos que no caben en el formato.
   - *Ejemplo*: Aproximar $23.4571$ a 2 decimales da $23.45$.
   - *Cota de error*: El error de truncamiento está acotado por la precisión del formato:
     $$\epsilon_{\text{trunc}} < P$$
2. **Redondeo (al más cercano)**: Consiste en elegir el número representable más cercano al número original.
   - *Ejemplo*: Aproximar $23.4571$ a 2 decimales da $23.46$ (puesto que el error $|23.4571 - 23.46| = 0.0029$ es menor que el de truncamiento $|23.4571 - 23.45| = 0.0071$).
   - *Cota de error*: El error de redondeo está acotado por la mitad de la precisión:
     $$\epsilon_{\text{red}} \le \frac{P}{2}$$
   - *Algoritmo de redondeo*: Una forma sencilla de redondear es sumar al número la mitad de la precisión y truncar el resultado:
     $$\hat{x}_{\text{red}} = \text{trunc}\left( x + \frac{P}{2} \right)$$
     *Ejemplo*: Redondear $1.526246$ con precisión $0.0001$:
     $$1.526246 + 0.00005 = 1.526296 \xrightarrow{\text{truncar a 4 decimales}} 1.5262$$

### 1.4 Desbordamientos: Overflow y Underflow
- **Desbordamiento a Infinito (Overflow)**: Ocurre cuando el resultado de una operación supera el límite superior de representación del formato (ej. sumar dos números grandes y exceder el rango).
- **Desbordamiento a Cero (Underflow)**: Ocurre cuando un número es distinto de cero pero su magnitud es menor que la precisión mínima del formato, representándose como cero. Esto causa graves inconvenientes, ya que invalida operaciones posteriores como la división (división por cero).


## 2. Sistemas de Numeración y Conversiones

### 2.1 Sistemas Posicionales de Base Fija
Un sistema de numeración es una metodología para representar valores. En un **sistema posicional**, el valor de cada dígito depende de su posición relativa al punto raíz. En base $\beta \ge 2$, un número real se expresa como:
$$n = (-1)^s \sum_{i=-\infty}^p d_i \beta^i = (-1)^s \left( d_p \beta^p + d_{p-1} \beta^{p-1} + \dots + d_0 \beta^0 + d_{-1} \beta^{-1} + \dots \right)$$
donde los dígitos satisfacen $0 \le d_i < \beta$. Las bases más comunes son $\beta=10$ (decimal), $\beta=2$ (binario), $\beta=8$ (octal) y $\beta=16$ (hexadecimal).

### 2.2 Conversión de Decimal a Base $b$: Divisiones Sucesivas
Para convertir la parte entera de un número decimal a base $b$, se divide sucesivamente entre $b$:
- El resto de la primera división da el dígito de menor peso ($d_0$).
- El cociente resultante se vuelve a dividir entre $b$, obteniendo el resto $d_1$.
- El proceso se repite hasta obtener un cociente cero. El último resto es el dígito más significativo ($d_p$).

**Ejemplo**: Convertir $1972_{10}$ a base 2, 8 y 16:
- **Base 8**:
  - $1972 / 8 = 246$ resto **4** ($d_0$)
  - $246 / 8 = 30$ resto **6** ($d_1$)
  - $30 / 8 = 3$ resto **6** ($d_2$)
  - $3 / 8 = 0$ resto **3** ($d_3$)
  $$\implies 1972_{10} = 3664_8$$
- **Base 16** (donde $10 \to A, 11 \to B, 12 \to C, 13 \to D, 14 \to E, 15 \to F$):
  - $1972 / 16 = 123$ resto **4** ($d_0$)
  - $123 / 16 = 7$ resto **11** $\to$ **B** ($d_1$)
  - $7 / 16 = 0$ resto **7** ($d_2$)
  $$\implies 1972_{10} = 7B4_{16}$$
  *Comprobación*: $7 \cdot 16^2 + 11 \cdot 16 + 4 = 1792 + 176 + 4 = 1972_{10}$.

### 2.3 Conversiones Directas entre Bases Potencia
Si $b_2 = b_1^k$, la conversión entre base $b_1$ y $b_2$ es directa agrupando de $k$ en $k$ dígitos a partir del punto raíz:
- **Base 2 a Base 8** ($8 = 2^3$, agrupar de 3 en 3):
  $$11110110100_2 = 011\ 110\ 110\ 100_2 \to 3664_8$$
- **Base 2 a Base 16** ($16 = 2^4$, agrupar de 4 en 4):
  $$11110110100_2 = 0111\ 1011\ 0100_2 \to 7B4_{16}$$
- **Parte Fraccionaria**: Se agrupa hacia la derecha a partir del punto decimal, completando con ceros a la derecha si es necesario:
  $$0.1_2 = 0.100_2 \to 0.4_8$$
  $$-11101.101_2 = 011\ 101.101_2 \to -35.5_8$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Flujograma visual del método de divisiones sucesivas: 1972 a Base 16
num = 1972
base = 16
hex_chars = "0123456789ABCDEF"

dividendos = []
cocientes = []
restos = []

curr = num
while True:
    div = curr // base
    rem = curr % base
    dividendos.append(curr)
    cocientes.append(div)
    restos.append(rem)
    if div == 0:
        break
    curr = div

fig, ax = plt.subplots(figsize=(10, 6))
plt.title('Diagrama del Método de Divisiones Sucesivas ($1972_{10} \\to 7B4_{16}$)', fontweight='bold', fontsize=14, color='navy')

# Dibujar cajas para cada división
for i in range(len(dividendos)):
    y_pos = 5.0 - i * 1.5
    # Caja Dividendo
    ax.text(1.0, y_pos, f"{dividendos[i]}", bbox=dict(facecolor='lightblue', alpha=0.5, boxstyle='round,pad=0.5'), fontsize=12, ha='center', va='center')
    # Flecha divisor
    ax.annotate(f"/ {base}", xy=(2.2, y_pos), xytext=(1.4, y_pos), arrowprops=dict(arrowstyle="->", color='gray', lw=1.5), fontsize=10, va='center')
    # Caja Cociente
    ax.text(2.6, y_pos, f"Cociente = {cocientes[i]}", bbox=dict(facecolor='lightgreen', alpha=0.5, boxstyle='round,pad=0.5'), fontsize=11, ha='center', va='center')
    # Resto
    rem_str = f"{restos[i]}" if restos[i] < 10 else f"{restos[i]} ({hex_chars[restos[i]]})"
    ax.text(5.0, y_pos, f"Resto = {rem_str}", bbox=dict(facecolor='lightcoral', alpha=0.5, boxstyle='round,pad=0.5'), fontsize=11, ha='center', va='center')

# Dibujar flecha de lectura de restos de abajo a arriba
ax.annotate('', xy=(6.5, 5.5), xytext=(6.5, 1.5), arrowprops=dict(arrowstyle="->", color='red', lw=2.5))
ax.text(6.8, 3.5, 'Sentido de Lectura\n(Menor a Mayor Peso)', rotation=270, ha='left', va='center', fontweight='bold', color='red')

ax.set_xlim(0.2, 8.2)
ax.set_ylim(1.0, 6.0)
ax.axis('off')
plt.show()
```

## 3. Operaciones Aritméticas en Base $b$ y Punto Flotante

### 3.1 Operaciones en Base 2
Las operaciones binarias siguen reglas análogas al sistema decimal, basándose en tablas de sumar y multiplicar reducidas:
- **Suma**: $0+0=0$, $0+1=1$, $1+1 = 10_2$ (escribimos 0 y acarreamos 1).
- **Producto**: $0 \cdot 0 = 0$, $0 \cdot 1 = 0$, $1 \cdot 1 = 1$.

**Ejemplos de la literatura**:
1. **Suma**: $10001_2 + 11011_2 = 101100_2$ ($17_{10} + 27_{10} = 44_{10}$).
2. **Producto**: $1101_2 \cdot 1110_2 = 10110110_2$ ($13_{10} \cdot 14_{10} = 182_{10}$).
3. **Resta**: $10100_2 - 1101_2 = 111_2$ ($20_{10} - 13_{10} = 7_{10}$).
4. **División**: $11100_2 / 1001_2 = 11_2$ con resto $1_2$ ($28_{10} / 9_{10} = 3_{10}$ con resto $1_{10}$).

### 3.2 Aritmética de Punto Flotante
Para realizar la suma y resta de dos números representados en punto flotante, se siguen los siguientes pasos sucesivos:
1. **Alineación de Mantisas**: Comparar exponentes. Tomar el número con menor exponente y desplazar su mantisa hacia la derecha (perdiendo dígitos significativos si excede la precisión) sumando 1 a su exponente por cada desplazamiento, hasta que los exponentes coincidan.
2. **Operación de Mantisas**: Sumar o restar algebraicamente las mantisas alineadas.
3. **Normalización**: Ajustar la mantisa del resultado para que el primer dígito significativo esté inmediatamente a la derecha del punto (o izquierda según formato) y corregir el exponente.
4. **Redondeo**: Descartar los bits sobrantes de la mantisa de acuerdo al criterio del formato:
   - **Truncamiento (Chopping)**: Cortar directamente.
   - **Redondeo de Von Neumann**: Si se descartan bits no nulos, se fuerza el bit menos significativo de la mantisa resultante a ser $1$. Esto evita sesgos y ahorra la propagación de acarreos de redondeo.
5. **Renormalización**: Realizar un último ajuste si el redondeo provocó desbordamiento de mantisa.


```python
# Simulación interactiva de operaciones en base 2 con validación en base 10

def bin_add(a_str, b_str):
    val_a = int(a_str, 2)
    val_b = int(b_str, 2)
    res = val_a + val_b
    return bin(res)[2:]

def bin_mul(a_str, b_str):
    val_a = int(a_str, 2)
    val_b = int(b_str, 2)
    res = val_a * val_b
    return bin(res)[2:]

a_bin, b_bin = '10001', '11011'  # 17 y 27
suma_bin = bin_add(a_bin, b_bin)
print(f"Suma Binaria:       {a_bin} + {b_bin} = {suma_bin} (Validación: {int(a_bin, 2)} + {int(b_bin, 2)} = {int(suma_bin, 2)})")

p_a, p_b = '1101', '1110'  # 13 y 14
prod_bin = bin_mul(p_a, p_b)
print(f"Producto Binario:   {p_a} * {p_b} = {prod_bin} (Validación: {int(p_a, 2)} * {int(p_b, 2)} = {int(prod_bin, 2)})")

r_a, r_b = '10100', '1101'  # 20 y 13
resta_bin = bin(int(r_a, 2) - int(r_b, 2))[2:]
print(f"Resta Binaria:      {r_a} - {r_b} = {resta_bin} (Validación: {int(r_a, 2)} - {int(r_b, 2)} = {int(resta_bin, 2)})")

d_a, d_b = '11100', '1001'  # 28 y 9
q_val = int(d_a, 2) // int(d_b, 2)
r_val = int(d_a, 2) % int(d_b, 2)
print(f"División Binaria:   {d_a} / {d_b} = {bin(q_val)[2:]} (Cociente), Resto = {bin(r_val)[2:]}")
print(f"                    (Validación: 28 / 9 = {q_val} con Resto {r_val})")
```

## 4. Propagación de Errores y Cancelación Substractiva

### 4.1 Cotas de Error en Operaciones Básicas
Si operamos con variables $x$ e $y$ aproximadas que poseen cotas de error absoluto $E_a(x)$ y $E_a(y)$, la propagación del error absoluto en las operaciones elementales está gobernada por las siguientes reglas límites de peor caso:
1. **Suma o Resta**: La cota de error absoluto del resultado es la suma de las cotas individuales:
   $$E_a(x \pm y) \le E_a(x) + E_a(y)$$
2. **Producto**: La cota de error relativo es la suma de las cotas relativas:
   $$E_r(x \cdot y) \le E_r(x) + E_r(y)$$
   El error absoluto estimado por diferenciales es:
   $$E_a(x \cdot y) \approx |x| E_a(y) + |y| E_a(x)$$
3. **Cociente**: La cota de error relativo es la suma de las cotas relativas:
   $$E_r(x / y) \le E_r(x) + E_r(y)$$
   El error absoluto estimado es:
   $$E_a(x / y) \approx \frac{1}{|y|} E_a(x) + \frac{|x|}{y^2} E_a(y)$$

### 4.2 Amplificación del Error en la Resta (Mal Condicionamiento)
Consideremos la resta de dos números aproximados $x$ e $y$ de magnitudes muy similares ($x \approx y$). Evaluamos la fórmula del error relativo para la diferencia:
$$E_r(x - y) = \frac{E_a(x-y)}{|x - y|} \le \frac{E_a(x) + E_a(y)}{|x - y|}$$
Expresado en términos de los errores relativos de cada operando:
$$E_r(x - y) \le \frac{|x|}{|x - y|} E_r(x) + \frac{|y|}{|x - y|} E_r(y)$$

El factor $\frac{|x|}{|x - y|}$ actúa como un **factor de amplificación o número de condición** de la operación:
- Si $x$ e $y$ son muy cercanos, el denominador $|x - y|$ tiende a cero.
- El factor de amplificación se dispara al infinito, multiplicando catastróficamente los errores relativos iniciales de los operandos.
- Esto demuestra de forma analítica rigurosa por qué restar números similares en computadoras destruye la significancia del resultado, convirtiéndose en una operación sumamente inestable y mal condicionada.

```python
# 1. Gráfica de la amplificación del error relativo en la resta
x_ref = 1.0
y_vals = np.linspace(0.8, 1.2, 500)
# Excluimos y = 1.0 para evitar la división por cero
y_vals = y_vals[np.abs(y_vals - x_ref) > 1e-4]

factor_amplificacion = x_ref / np.abs(x_ref - y_vals)

fig, axs = plt.subplots(1, 2, figsize=(14, 6.5))

axs[0].plot(y_vals, factor_amplificacion, 'r-', linewidth=2.0, label='Factor $\\frac{|x|}{|x-y|}$')
axs[0].set_yscale('log')
axs[0].axvline(1.0, color='black', linestyle=':', label='Divergencia en $y = x$')
axs[0].set_title('Factor de Amplificación del Error en la Resta ($x=1.0$)', fontweight='bold', color='navy')
axs[0].set_xlabel('Valor de $y$')
axs[0].set_ylabel('Factor de amplificación (Escala Logarítmica)')
axs[0].legend()

# 2. Simulación de Monte Carlo para la propagación del error en el producto
# x_real = 10.0, e_x = +- 0.1
# y_real = 5.0,  e_y = +- 0.05
n_samples = 10000
x_real, y_real = 10.0, 5.0
dx, dy = 0.1, 0.05

x_samples = np.random.uniform(x_real - dx, x_real + dx, n_samples)
y_samples = np.random.uniform(y_real - dy, y_real + dy, n_samples)

real_product = x_real * y_real
product_samples = x_samples * y_samples
error_samples = np.abs(product_samples - real_product)

# Cota teórica absoluta: E_a(xy) <= |x|E_a(y) + |y|E_a(x) = 10*(0.05) + 5*(0.1) = 0.5 + 0.5 = 1.0
cota_teorica = x_real * dy + y_real * dx

axs[1].hist(error_samples, bins=50, color='mediumseagreen', alpha=0.7, edgecolor='black', label='Errores simulados')
axs[1].axvline(cota_teorica, color='red', linestyle='--', linewidth=2.5, label=f'Cota Teórica Absoluta = {cota_teorica:.2f}')
axs[1].set_title('Propagación de Error en el Producto (Simulación Monte Carlo)', fontweight='bold', color='navy')
axs[1].set_xlabel('Error absoluto real $|xy - x_{aprox}y_{aprox}|$')
axs[1].set_ylabel('Frecuencia')
axs[1].legend()

plt.tight_layout()
plt.show()
```

## 5. Resumen de la Lección

1. **Representación de la Información**: Condicionada por limitaciones físicas de la memoria. Define rango (límites) y precisión (paso mínimo).
2. **Aproximación**: El truncamiento descarta decimales directamente (error máximo = precisión $P$). El redondeo aproxima al más cercano, reduciendo el error máximo a $P/2$.
3. **Desbordamientos**: Overflow denota rebasar el valor representable más alto; underflow denota caer por debajo de la precisión representable (desbordamiento a cero).
4. **Sistemas Numéricos**: Los sistemas basados en raíces posicionales (binario, octal, hexadecimal) permiten manipular información computacional de forma ágil.
5. **Conversión**: Se realiza por divisiones sucesivas para partes enteras (o multiplicaciones sucesivas para fraccionarias) o agrupando de $k$ en $k$ dígitos en bases potencia ($b_2 = b_1^k$).
6. **Aritmética de Punto Flotante**: Sigue etapas lógicas de alineación de mantisas, operación, normalización y redondeo (destacando el redondeo de Von Neumann para evitar sesgo de arrastre).
7. **Propagación del Error**: El error absoluto de sumas y restas es acotado por la suma de errores individuales. El error relativo de la resta de valores muy similares explota numéricamente debido al factor de amplificación $\frac{|x|}{|x-y|}$.

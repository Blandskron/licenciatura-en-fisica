## 1. Divisibilidad y Aritmética Modular

### 1.1 Divisibilidad en los Enteros ($\mathbb{Z}$)

De acuerdo con Kenneth [1], si $a$ y $b$ son números enteros con $a \neq 0$, decimos que **$a$ divide a $b$** si existe un entero $c$ tal que:
$$b = a \cdot c$$
* Se denota por $a \mid b$ (se lee *"a divide a b"*).
* Si $a$ no divide a $b$, se denota por $a \nmid b$.
* Cuando $a \mid b$, decimos que $a$ es un **factor** o **divisor** de $b$, y que $b$ es un **múltiplo** de $a$.

#### Propiedades Formales de la Divisibilidad (Zaldívar [2])
Para enteros cualesquiera $a, b, c, r, s$ (con divisores no nulos):
1. **Reflexividad:** $a \mid a$.
2. **Transitividad:** Si $a \mid b$ y $b \mid c$, entonces $a \mid c$.
3. **Elemento Identidad:** $1 \mid a$ para todo $a \in \mathbb{Z}$.
4. **División del Cero:** $a \mid 0$ para todo $a \neq 0$.
5. **Multiplicación:** Si $a \mid b$, entonces $a \mid b \cdot r$ para cualquier entero $r$.
6. **Aditividad:** Si $a \mid b$ y $a \mid c$, entonces $a \mid (b + c)$.
7. **Combinación Lineal:** Si $a \mid b$ y $a \mid c$, entonces $a \mid (b \cdot r + c \cdot s)$ para cualquier par de enteros $r, s$.
8. **Simetría y Magnitud:** Si $a \mid b$ y $b \mid a$, entonces $a = \pm b$.

### 1.2 El Algoritmo de la División

A pesar de su nombre histórico, el **algoritmo de la división** es un teorema que establece la existencia y unicidad de la división entera.

* **Teorema:** Sean $a$ un entero y $d$ un entero positivo. Existen dos únicos enteros $q$ (cociente) y $r$ (resto) tales que:
  $$a = d \cdot q + r \quad \text{con} \quad 0 \le r < d$$
* **Notación:**
  * Cociente: $q = a \text{ div } d = \lfloor a / d \rfloor$
  * Resto (Residuo): $r = a \bmod d = a - d \cdot \lfloor a / d \rfloor$

### 1.3 Aritmética Modular

La **congruencia modular** agrupa los números enteros según el residuo que dejan al dividirse por un entero positivo $m$ (llamado **módulo**).

* **Definición:** Si $a$ y $b$ son enteros y $m$ es un entero positivo, decimos que **$a$ es congruente con $b$ módulo $m$** si $m$ divide a la diferencia $a - b$. Se denota por:
  $$a \equiv b \pmod m$$
* **Teoremas Clave (Kenneth [1], Mora [3]):**
  1. $a \equiv b \pmod m$ si y solo si $a \bmod m = b \bmod m$.
  2. $a \equiv b \pmod m$ si y solo si existe un entero $k$ tal que $a = b + k \cdot m$.
  3. **Compatibilidad Algebraica:** Si $a \equiv b \pmod m$ y $c \equiv d \pmod m$, entonces:
     $$(a + c) \equiv (b + d) \pmod m$$
     $$(a \cdot c) \equiv (b \cdot d) \pmod m$$


```python
# 1. Implementación del Algoritmo de la División
def division_entera(a, d):
    # En Python, // es el cociente (floor division) y % es el módulo
    q = a // d
    r = a % d
    return q, r

a, d = 27, 4
q, r = division_entera(a, d)
print(f"Algoritmo de la division para a={a}, d={d}:")
print(f"  Cociente (q = a div d) = {q}")
print(f"  Resto (r = a mod d) = {r}")
print(f"  Verificacion (a == d*q + r): {a} == {d}*{q} + {r} -> {a == d*q + r}")
print("-" * 50)

# 2. Operaciones Modulares y Compatibilidad
m = 7
x1, x2 = 15, 22
y1, y2 = 3, 10  # Note que 15 == 1 (mod 7) y 22 == 1 (mod 7), 3 == 3 (mod 7) y 10 == 3 (mod 7)

print(f"Compatibilidad Modular modulo m={m}:")
print(f"  ¿15 es congruente con 22 mod 7? {x1 % m == x2 % m}")
print(f"  ¿3 es congruente con 10 mod 7? {y1 % m == y2 % m}")

# Evaluamos la compatibilidad de la suma
suma_original = (x1 + y1) % m
suma_congruente = (x2 + y2) % m
print(f"  Suma mod 7: (15+3) mod 7 = {suma_original} == (22+10) mod 7 = {suma_congruente} -> {suma_original == suma_congruente}")

# Evaluamos la compatibilidad del producto
prod_original = (x1 * y1) % m
prod_congruente = (x2 * y2) % m
print(f"  Producto mod 7: (15*3) mod 7 = {prod_original} == (22*10) mod 7 = {prod_congruente} -> {prod_original == prod_congruente}")

```

## 2. Números Primos y Factorización

Los números primos son los ladrillos de construcción de los enteros.

### 2.1 Definición y Propiedades
* **Número Primo:** Un entero positivo $p > 1$ cuyos únicos divisores positivos son 1 y $p$.
* **Número Compuesto:** Un entero positivo $n > 1$ que no es primo (tiene divisores distintos de 1 y $n$).
* *(Nota: El número 1 no se considera primo ni compuesto por convención algebraica).*

### 2.2 Criterio de Primalidad de la Raíz Cuadrada
* **Teorema:** Si $n$ es un número entero compuesto, entonces $n$ tiene un divisor primo $p$ tal que $p \le \sqrt{n}$.
* **Corolario:** Para probar si un número $n > 1$ es primo, basta con verificar si es divisible por algún número primo menor o igual a $\sqrt{n}$. Si no tiene divisores menores o iguales a $\sqrt{n}$, entonces $n$ es primo.
  * *Ejemplo:* ¿Es 53 primo?
    $$\sqrt{53} \approx 7.28$$
    Los primos menores o iguales a 7.28 son $\{2, 3, 5, 7\}$. Como ninguno de estos primos divide a 53, entonces 53 es primo.

### 2.3 Criba de Eratóstenes
Es un algoritmo eficiente para hallar todos los números primos menores o iguales a un entero dado $n$:
1. Escribir una lista de enteros del 2 al $n$.
2. Tomar el primer número no tachado (el 2) como primo y tachar todos sus múltiplos.
3. Tomar el siguiente número no tachado (el 3) como primo y tachar sus múltiplos.
4. Repetir hasta que el número de prueba sea superior a $\sqrt{n}$. Los números restantes no tachados son los primos.

### 2.4 Teorema Fundamental de la Aritmética
* Todo entero positivo mayor que 1 puede ser escrito de forma única como un número primo o como el producto de dos o más primos, donde los factores primos se escriben en orden no decreciente.
  $$n = p_1^{a_1} \cdot p_2^{a_2} \cdots p_k^{a_k}$$
  *Ejemplo:* $40 = 2^3 \cdot 5^1$. Descomposición en factores primos.


```python
# 1. Test de Primalidad optimizado hasta sqrt(n)
def es_primo(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Probamos divisores de la forma 6k +/- 1 hasta sqrt(n)
    limite = int(n**0.5)
    i = 5
    while i <= limite:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# 2. Criba de Eratóstenes para obtener primos hasta n
def criba_eratostenes(n):
    primo = [True] * (n + 1)
    primo[0] = primo[1] = False
    
    p = 2
    while (p * p <= n):
        if (primo[p] == True):
            # Tachar los múltiplos de p
            for i in range(p * p, n + 1, p):
                primo[i] = False
        p += 1
    
    # Recolectar primos
    return [i for i in range(2, n + 1) if primo[i]]

# 3. Factorización en Factores Primos
def factorizar_primo(n):
    factores = {}
    d = 2
    temp = n
    while d * d <= temp:
        while (temp % d) == 0:
            factores[d] = factores.get(d, 0) + 1
            temp //= d
        d += 1
    if temp > 1:
        factores[temp] = 1
    return factores

# Probamos las funciones
num_test = 53
print(f"¿{num_test} es primo? {es_primo(num_test)}")
print("Primos menores a 50 usando la Criba:", criba_eratostenes(50))

num_fact = 40
factores = factorizar_primo(num_fact)
factores_str = " * ".join(f"{p}^{exp}" for p, exp in factores.items())
print(f"Factorizacion prima de {num_fact}: {factores_str}")

```

## 3. MCD, mcm y Algoritmo de Euclides

### 3.1 Definición
* **Máximo Común Divisor ($	ext{mcd}(a, b)$):** El entero más grande que divide tanto a $a$ como a $b$.
* **Mínimo Común Múltiplo ($	ext{mcm}(a, b)$):** El menor entero positivo divisible tanto por $a$ como por $b$.
* **Primos Relativos (Coprimos):** Dos enteros $a$ y $b$ son coprimos si y solo si $	ext{mcd}(a, b) = 1$.

### 3.2 Relación Fundamental
$$\text{mcd}(a, b) \cdot \text{mcm}(a, b) = a \cdot b \implies \text{mcm}(a, b) = \frac{a \cdot b}{\text{mcd}(a, b)}$$

### 3.3 El Algoritmo de Euclides

Es el método más eficiente para calcular el MCD de dos números enteros, basado en el principio del residuo modular.

* **Principio:** Si $a = b \cdot q + r$, entonces:
  $$\text{mcd}(a, b) = \text{mcd}(b, r) = \text{mcd}(b, a \bmod b)$$
  El proceso se repite dividiendo sucesivamente el divisor por el resto anterior hasta que el resto es cero. El último divisor distinto de cero es el MCD.

#### Ejemplo del Algoritmo de Euclides: Calcular $\text{mcd}(287, 91)$
1. $287 = 91 \cdot 3 + 14 \implies \text{mcd}(287, 91) = \text{mcd}(91, 14)$
2. $91 = 14 \cdot 6 + 7 \implies \text{mcd}(91, 14) = \text{mcd}(14, 7)$
3. $14 = 7 \cdot 2 + 0 \implies \text{mcd}(14, 7) = 7$
Por lo tanto, el MCD es 7.

### 3.4 Algoritmo de Euclides Extendido
El algoritmo de Euclides extendido no solo calcula el MCD $g$, sino que también halla dos enteros coeficientes $s$ y $t$ (coeficientes de Bezout) tales que:
$$a \cdot s + b \cdot t = \text{mcd}(a, b)$$
Esta identidad es clave en la resolución de congruencias lineales y en la búsqueda de inversos modulares.


```python
# 1. Algoritmo de Euclides simple (MCD)
def mcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# 2. Mínimo Común Múltiplo (mcm)
def mcm(a, b):
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // mcd(a, b)

# 3. Algoritmo de Euclides Extendido
def euclides_extendido(a, b):
    # Retorna (mcd, s, t) tal que a*s + b*t = mcd
    if a == 0:
        return b, 0, 1
    mcd_val, s1, t1 = euclides_extendido(b % a, a)
    
    # Actualizamos los coeficientes
    s = t1 - (b // a) * s1
    t = s1
    return mcd_val, s, t

# Probamos con el ejemplo del PDF: mcd(287, 91)
val_a, val_b = 287, 91
gcd = mcd(val_a, val_b)
print(f"MCD({val_a}, {val_b}) = {gcd}")
print(f"mcm({val_a}, {val_b}) = {mcm(val_a, val_b)}")

# Probamos el extendido
g, s, t = euclides_extendido(val_a, val_b)
print(f"Euclides Extendido para {val_a} y {val_b}:")
print(f"  MCD = {g}")
print(f"  Coeficientes Bezout: s = {s}, t = {t}")
print(f"  Verificacion: {val_a}*({s}) + {val_b}*({t}) == {val_a*s + val_b*t}")

```

## 4. Congruencias Lineales y Teorema Chino del Resto

### 4.1 Congruencias Lineales y el Inverso Modular
Una congruencia lineal es una ecuación de la forma:
$$a \cdot x \equiv b \pmod m$$
Donde $x$ es la variable entera a despejar. Para resolverla, necesitamos el concepto de **inverso de $a$ módulo $m$** (denotado por $\bar{a}$):
$$a \cdot \bar{a} \equiv 1 \pmod m$$

* **Teorema de Existencia:** El inverso modular $\bar{a}$ existe y es único módulo $m$ si y solo si $a$ y $m$ son coprimos:
  $$\text{mcd}(a, m) = 1$$
* **Cálculo del Inverso:** Utilizando el Algoritmo de Euclides Extendido en $a$ y $m$:
  $$a \cdot s + m \cdot t = 1 \implies a \cdot s \equiv 1 \pmod m$$
  Por lo tanto, $\bar{a} = s \bmod m$ es el inverso modular.
* **Resolución de la Congruencia:** Multiplicamos ambos miembros por $\bar{a}$:
  $$\bar{a} \cdot a \cdot x \equiv \bar{a} \cdot b \pmod m \implies x \equiv \bar{a} \cdot b \pmod m$$

### 4.2 El Teorema Chino del Resto (CRT)
Permite resolver sistemas de congruencias simultáneas de la forma:
$$x \equiv a_1 \pmod{m_1}$$
$$x \equiv a_2 \pmod{m_2}$$
$$\vdots$$
$$x \equiv a_k \pmod{m_k}$$
Si los módulos $m_i$ son primos relativos dos a dos (coprimos), entonces el sistema tiene una **solución única módulo $M = m_1 \cdot m_2 \cdots m_k$**.

#### Ejemplo del PDF (Juego de Naipes)
Buscamos un entero $n$ tal que:
1. $n \equiv 1 \pmod 7$
2. $n \equiv 10 \pmod{11}$

* **Resolución Analítica:**
  * De (1): $n = 1 + 7s$.
  * Sustituimos en (2): $1 + 7s \equiv 10 \implies 7s \equiv 9 \pmod{11}$.
  * Buscamos el inverso de 7 módulo 11, el cual es 8 (ya que $7 \cdot 8 = 56 \equiv 1 \pmod{11}$).
  * Multiplicamos por 8: $s \equiv 72 \equiv 6 \pmod{11} \implies s = 6 + 11t$.
  * Sustituimos $s$ en la primera ecuación: $n = 1 + 7(6 + 11t) = 43 + 77t$.
  * La solución general es $n \equiv 43 \pmod{77}$. La mínima cantidad positiva de cartas es $43$.


```python
# 1. Inverso Modular usando Euclides Extendido
def inverso_modular(a, m):
    g, s, t = euclides_extendido(a, m)
    if g != 1:
        raise ValueError(f"El inverso modular no existe porque MCD({a}, {m}) = {g} != 1")
    return s % m

# 2. Resolver congruencia lineal a*x == b (mod m)
def resolver_congruencia_lineal(a, b, m):
    # Buscamos mcd(a, m)
    g = mcd(a, m)
    if b % g != 0:
        return None # No tiene solución
    
    # Reducimos la congruencia dividiendo entre el MCD
    a_red = a // g
    b_red = b // g
    m_red = m // g
    
    # Calculamos el inverso de a_red mod m_red
    inv = inverso_modular(a_red, m_red)
    sol_base = (inv * b_red) % m_red
    
    # Retornamos todas las soluciones mod m
    return [sol_base + i * m_red for i in range(g)]

# 3. Solver general del Teorema Chino del Resto (CRT)
def teorema_chino_resto(a_list, m_list):
    # a_list: [a_1, a_2, ...], m_list: [m_1, m_2, ...]
    # Primero calculamos el producto de todos los módulos M
    M = 1
    for m in m_list:
        M *= m
        
    x = 0
    for a_i, m_i in zip(a_list, m_list):
        M_i = M // m_i
        # Calculamos el inverso de M_i modulo m_i
        inv_i = inverso_modular(M_i, m_i)
        x += a_i * M_i * inv_i
        
    return x % M, M

# Verificaciones
# Inverso de 3 mod 7
print("Inverso de 3 mod 7:", inverso_modular(3, 7))

# Resolver 3x == 4 (mod 7)
sol = resolver_congruencia_lineal(3, 4, 7)
print("Soluciones de 3x == 4 (mod 7):", sol)

# Resolver el acertijo del PDF: n == 1 (mod 7) y n == 10 (mod 11)
n_sol, M_total = teorema_chino_resto([1, 10], [7, 11])
print(f"Solucion del acertijo lógico (naipes): n == {n_sol} (mod {M_total})")

```

## 5. Teoremas de Fermat, Euler, Raíces Primitivas y Logaritmo Discreto

### 5.1 El Pequeño Teorema de Fermat (FLT)
* **Teorema:** Si $p$ es primo y $a$ es un entero no divisible por $p$, entonces:
  $$a^{p-1} \equiv 1 \pmod p$$
* **Pseudoprimos:** Enteros compuestos $n$ tales que $2^{n-1} \equiv 1 \pmod n$. Actúan como "falsos primos" en algunas pruebas rápidas de primalidad.
* **Números de Carmichael:** Enteros compuestos $n$ que son pseudoprimos para **cualquier base $b$** coprima con $n$. El número de Carmichael más pequeño es $561$.

### 5.2 La Función $\varphi(n)$ de Euler y el Teorema de Euler
* **Función Totiente de Euler $\varphi(n)$:** Cantidad de enteros positivos menores que $n$ que son coprimos con $n$.
  * Si $p$ es primo: $\varphi(p) = p - 1$.
  * Si $n = p^k$: $\varphi(p^k) = p^k - p^{k-1}$.
* **Teorema de Euler:** Si $a$ y $m$ son coprimos, entonces:
  $$a^{\varphi(m)} \equiv 1 \pmod m$$

### 5.3 Raíces Primitivas y Logaritmo Discreto
* **Orden de $a$ módulo $n$ ($\text{ord}_n(a)$):** El menor entero positivo $k$ tal que $a^k \equiv 1 \pmod n$.
* **Raíz Primitiva:** Un entero $g$ es una raíz primitiva módulo $n$ si el orden de $g$ es exactamente igual a $\varphi(n)$.
  * Si $p$ es primo y $g$ es raíz primitiva de $p$, entonces las potencias consecutivas $g, g^2, g^3, \dots, g^{p-1} \pmod p$ generan todos los elementos del grupo multiplicativo $\mathbb{Z}_p^*$ (es decir, todos los residuos coprimos con $p$).
* **Logaritmo Discreto:** Si $g$ es raíz primitiva módulo $p$ y $a \equiv g^x \pmod p$, entonces $x$ es el **logaritmo discreto** de $a$ en base $g$ módulo $p$, denotado por:
  $$x = \text{Ind}_g(a) \quad \text{o} \quad x = \log_g a \pmod p$$
  *(Nota: A diferencia del logaritmo real, calcular logaritmos discretos modulares es un problema computacionalmente intratable para módulos grandes, lo que permite la seguridad de criptosistemas como Diffie-Hellman).*


```python
# 1. Función Totiente de Euler
def phi(n):
    resultado = n
    p = 2
    temp = n
    while p * p <= temp:
        if temp % p == 0:
            while temp % p == 0:
                temp //= p
            resultado -= resultado // p
        p += 1
    if temp > 1:
        resultado -= resultado // temp
    return resultado

# 2. Verificación de Fermat y Pseudoprimalidad para el número de Carmichael 561
carmichael = 561
print(f"¿{carmichael} es primo? {es_primo(carmichael)}")
# Comprobamos si es pseudoprimo para base b=2: 2^(n-1) == 1 (mod n)
es_pseudoprimo_2 = pow(2, carmichael - 1, carmichael) == 1
print(f"¿561 cumple 2^(560) == 1 (mod 561)? {es_pseudoprimo_2}")

# 3. Encontrar Raíces Primitivas modulo p (p primo)
def encontrar_raices_primitivas(p):
    if not es_primo(p):
        return []
    phi_p = p - 1
    # Buscamos los factores primos de phi(p)
    factores = list(factorizar_primo(phi_p).keys())
    
    raices = []
    for g in range(2, p):
        # g es raíz primitiva si para todo factor q de p-1: g^((p-1)/q) != 1 (mod p)
        es_raiz = True
        for q in factores:
            if pow(g, phi_p // q, p) == 1:
                es_raiz = False
                break
        if es_raiz:
            raices.append(g)
    return raices

# 4. Logaritmo Discreto (Fuerza Bruta para números pequeños)
def logaritmo_discreto(g, a, p):
    # Resuelve g^x == a (mod p)
    for x in range(p):
        if pow(g, x, p) == a:
            return x
    return None

p_primo = 5
print(f"\nFuncion phi({p_primo}) = {phi(p_primo)}")
raices = encontrar_raices_primitivas(p_primo)
print(f"Raices primitivas modulo {p_primo}: {raices}")

# Logaritmo discreto: resolver 3^x == 2 (mod 5)
g_primitivo = 3
obj = 2
dl = logaritmo_discreto(g_primitivo, obj, p_primo)
print(f"Logaritmo discreto de {obj} en base {g_primitivo} mod {p_primo}: x = {dl}")
print(f"  Verificacion: {g_primitivo}^{dl} mod {p_primo} == {pow(g_primitivo, dl, p_primo)}")

```

## 6. Algoritmo de Intercambio de Claves de Diffie-Hellman

El intercambio de claves de **Diffie-Hellman** es un protocolo criptográfico que permite a dos partes acordar de forma segura una **clave secreta común** sobre un canal de comunicación público no seguro.

### 6.1 Pasos del Protocolo

Supongamos que **Helen** y **Dan** quieren acordar una clave compartida:

1. **Parámetros Públicos:** Acuerdan un número primo grande $p$ y una raíz primitiva $g$ de $p$. Estos valores son públicos y conocidos por cualquiera.
2. **Claves Privadas:**
   * Helen elige un número entero secreto $n$.
   * Dan elige un número entero secreto $m$.
3. **Claves Públicas (Intercambio):**
   * Helen calcula su clave pública $A = g^n \bmod p$ y se la envía a Dan.
   * Dan calcula su clave pública $B = g^m \bmod p$ y se la envía a Helen.
4. **Cálculo del Secreto Compartido ($s$):**
   * Helen recibe $B$ y calcula $s = B^n \bmod p$.
   * Dan recibe $A$ y calcula $s = A^m \bmod p$.

#### Consistencia Matemática:
Ambos obtienen exactamente el mismo valor de clave secreta $s$, ya que:
$$s = B^n \equiv (g^m)^n \equiv g^{mn} \equiv (g^n)^m \equiv A^m \pmod p$$

### 6.2 Seguridad Lógica
Un atacante pasivo intercepta el canal de comunicación y conoce $p, g, A, B$. Para descifrar el secreto $s = g^{mn} \bmod p$, el atacante debe resolver los exponentes $n$ o $m$ a partir de las potencias públicas:
$$A \equiv g^n \pmod p \implies n = \log_g A \pmod p$$
Este es el **Problema del Logaritmo Discreto (DLP)**. Para primos de gran magnitud (ej. más de 2048 bits), este problema es matemáticamente intratable utilizando computadoras tradicionales, lo que garantiza la seguridad del cifrado.


```python
# Simulación del Intercambio de Claves Diffie-Hellman

# 1. Parametros Públicos (Primo p y raíz primitiva g de p)
p = 997  # Primo de prueba
g = 129  # Raíz primitiva modulo p (verificada)

print("--- PARAMETROS PUBLICOS ---")
print(f"Primo p: {p}")
print(f"Raiz primitiva g: {g}")
print("-" * 50)

# 2. Claves Privadas (Secretas)
n_Helen = 456  # Clave privada de Helen
m_Dan = 789    # Clave privada de Dan

print("--- CLAVES PRIVADAS ---")
print(f"Clave privada de Helen (secreta n): {n_Helen}")
print(f"Clave privada de Dan (secreta m): {m_Dan}")
print("-" * 50)

# 3. Cálculo de Claves Públicas (Intercambiadas)
# A = g^n mod p
A_Helen = pow(g, n_Helen, p)
# B = g^m mod p
B_Dan = pow(g, m_Dan, p)

print("--- CLAVES PUBLICAS (ENVIADAS POR EL CANAL) ---")
print(f"Clave publica de Helen (A = g^n mod p): {A_Helen}")
print(f"Clave publica de Dan (B = g^m mod p): {B_Dan}")
print("-" * 50)

# 4. Cálculo del Secreto Compartido por ambas partes
secreto_Helen = pow(B_Dan, n_Helen, p)
secreto_Dan = pow(A_Helen, m_Dan, p)

print("--- SECRETO COMPARTIDO ---")
print(f"Secreto calculado por Helen: {secreto_Helen}")
print(f"Secreto calculado por Dan: {secreto_Dan}")
print(f"¿Los secretos coinciden? {secreto_Helen == secreto_Dan}")
print("-" * 50)

# 5. Ataque de Interceptacion por Fuerza Bruta (Simulación de un atacante)
print("--- ATAQUE DE INTERCEPTACION ---")
# El atacante intercepta A y B, y conoce g y p. Intentará resolver el logaritmo discreto.
log_Helen = logaritmo_discreto(g, A_Helen, p)
log_Dan = logaritmo_discreto(g, B_Dan, p)

print(f"Atacante descubre clave privada de Helen (n): {log_Helen}")
print(f"Atacante descubre clave privada de Dan (m): {log_Dan}")
clave_descifrada = pow(g, log_Helen * log_Dan, p)
print(f"Atacante descifra la clave secreta compartida (s): {clave_descifrada}")

```

## 7. Resumen de la Lección

1. **Divisibilidad:** Estudia la estructura de factores y múltiplos de los enteros, caracterizados por el algoritmo de la división.
2. **Aritmética Modular:** Permite operar con clases de residuos bajo un módulo $m$, preservando la consistencia algebraica para sumas y productos.
3. **MCD y mcm:** El algoritmo de Euclides proporciona un método rápido para calcular el MCD, mientras que su versión extendida calcula los coeficientes de Bezout para hallar inversos modulares.
4. **Congruencias Lineales y CRT:** Las ecuaciones modulares se resuelven mediante inversos modulares, y el Teorema Chino del Resto unifica múltiples congruencias independientes en una solución única.
5. **Teoremas Clásicos:** El Pequeño Teorema de Fermat y el Teorema de Euler definen propiedades cíclicas fundamentales y extienden el concepto de orden y raíces primitivas modulo $n$.
6. **Criptografía:** El algoritmo Diffie-Hellman aprovecha la asimetría computacional entre la exponenciación modular (rápida) y el logaritmo discreto (extremadamente lento y costoso) para acordar secretos en canales públicos.

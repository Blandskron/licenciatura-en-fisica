## 1. Introducción a los Árboles y Terminología

### 1.1 Definición Matemática
* **Árbol:** Un grafo no dirigido, simple, conexo y sin ciclos (acíclico).
* **Bosque (Forest):** Un grafo no dirigido y acíclico cuyas componentes conexas son árboles individuales.
* **Teorema de Unicidad de Caminos:** Un grafo no dirigido $G$ es un árbol si y solo si existe un **único camino simple** entre cualquier pareja de vértices en $G$.

### 1.2 Árbol con Raíz (Rooted Tree)
Un **árbol con raíz** es un árbol en el cual uno de sus vértices ha sido designado como la **raíz**. Las aristas adquieren una orientación natural (alejándose o acercándose a la raíz).
* **Padre (Parent):** Si existe una arista dirigida de $u \to v$, entonces $u$ es el padre de $v$.
* **Hijo (Child):** Si $u$ es el padre de $v$, entonces $v$ es el hijo de $u$.
* **Hermanos (Siblings):** Vértices que comparten el mismo padre.
* **Antecesores (Ancestors):** Vértices en el camino único desde la raíz hasta el nodo (excluyendo al nodo mismo).
* **Descendientes (Descendants):** Vértices que tienen al nodo como antecesor.
* **Hoja (Leaf / Nodo Terminal):** Vértices que no tienen hijos (en árboles con raíz, su grado de salida es 0).
* **Vértice interno (Internal Node):** Vértices que tienen al menos un hijo.
* **Subárbol:** Subgrafo que consiste en un vértice, todos sus descendientes y las aristas asociadas.

### 1.3 Propiedades Métricas
* **Nivel:** La longitud del camino único desde la raíz hasta el nodo más uno. La raíz tiene nivel 1 (según la convención del PDF).
* **Altura ($h$):** El nivel máximo de cualquier nodo en el árbol.
* **Peso:** El número total de nodos en el árbol ($n$).
* **Grado de un Nodo:** En árboles con raíz, es el número de hijos que tiene.
* **Grado del Árbol:** El grado máximo de cualquiera de sus nodos.
* **Árbol m-ario:** Árbol con raíz donde cada vértice interno tiene como máximo $m$ hijos. Si $m=2$, se denomina **árbol binario**.
* **Árbol m-ario completo:** Aquel donde cada vértice interno tiene **exactamente** $m$ hijos.


```python
class SimpleNode:
    def __init__(self, value):
        self.value = value
        self.children = []
        
    def add_child(self, child_node):
        self.children.append(child_node)

def print_tree(node, level=0):
    indent = "    " * level
    print(f"{indent}* Node: {node.value} (Nivel {level + 1})")
    for child in node.children:
        print_tree(child, level + 1)

# Construir árbol sencillo de la Figura 1 del PDF
# Estructura: raíz 4 con hijos 1, 2, 3, 5. Nodos 1 y 2 son hojas. 3 tiene hijo 6. 5 tiene hijo 6? No, según la fig 1:
# Raíz es 4. Conecta a 1, 2, 3, 5. 5 conecta a 6.
raiz = SimpleNode(4)
nodo1 = SimpleNode(1)
nodo2 = SimpleNode(2)
nodo3 = SimpleNode(3)
nodo5 = SimpleNode(5)
nodo6 = SimpleNode(6)

raiz.add_child(nodo1)
raiz.add_child(nodo2)
raiz.add_child(nodo3)
raiz.add_child(nodo5)
nodo5.add_child(nodo6)

print("--- Representación Estructural del Árbol (Figura 1) ---")
print_tree(raiz)

```

## 2. Propiedades Matemáticas de los Árboles

El análisis estructural de los árboles permite derivar propiedades cuantitativas exactas mediante relaciones algebraicas:

### 2.1 Teorema Fundamental de Aristas
Un árbol con $n$ vértices tiene exactamente **$n-1$ aristas**.
* **Demostración por inducción:** Si $n=1$, tiene 0 aristas ($1-1=0$, base cierta). Asumiendo que se cumple para $k$ vértices, si añadimos un nuevo vértice conexo sin ciclos, debemos agregar exactamente una única arista para unirlo a la estructura existente, resultando en $k$ aristas para $k+1$ vértices.

### 2.2 Relaciones en Árboles m-arios Completos
Sea $n$ el número total de vértices, $i$ el número de vértices internos y $l$ el número de hojas. Para un árbol $m$-ario completo se cumplen de forma exacta las siguientes ecuaciones:

1. El número total de vértices en función de los internos:
   $$n = m \cdot i + 1$$
2. Dado el número total de vértices $n$:
   - Vértices internos: $i = \frac{n-1}{m}$
   - Hojas: $l = \frac{(m-1)n + 1}{m}$
3. Dado el número de internos $i$:
   - Vértices: $n = m \cdot i + 1$
   - Hojas: $l = (m-1)i + 1$
4. Dado el número de hojas $l$:
   - Vértices: $n = \frac{m \cdot l - 1}{m - 1}$
   - Vértices internos: $i = \frac{l-1}{m-1}$

### 2.3 Teorema de Cota en Hojas y Altura
* Un árbol $m$-ario de altura $h$ tiene como máximo $m^{h-1}$ hojas (si la raíz es nivel 1, o $m^h$ si la raíz es nivel 0). Según el PDF:
  $$l \le m^h \quad (\text{hojas a lo sumo})$$
* **Corolario de Altura:** Si un árbol $m$-ario de altura $h$ tiene $l$ hojas, entonces:
  $$h \ge \lceil \log_m l \rceil$$
  Si el árbol es completo y equilibrado (todas las hojas están en el último o penúltimo nivel), entonces la altura es exactamente:
  $$h = \lceil \log_m l \rceil$$


```python
import math

def calcular_propiedades_m_ario(m, n=None, i=None, l=None):
    """
    Calcula las propiedades estructurales de un árbol m-ario completo.
    """
    if n is not None:
        i_calc = (n - 1) // m
        l_calc = ((m - 1) * n + 1) // m
        return n, i_calc, l_calc
    elif i is not None:
        n_calc = m * i + 1
        l_calc = (m - 1) * i + 1
        return n_calc, i, l_calc
    elif l is not None:
        n_calc = (m * l - 1) // (m - 1)
        i_calc = (l - 1) // (m - 1)
        return n_calc, i_calc, l
    else:
        raise ValueError("Se debe proporcionar al menos uno de: n, i, l")

# Probar propiedades para un árbol binario completo (m = 2) con 9 hojas (l = 9)
m = 2
l_val = 9
n, i, l = calcular_propiedades_m_ario(m, l=l_val)
h_min = math.ceil(math.log(l, m))

print(f"--- Análisis Métrico del Árbol {m}-ario Completo (l = {l_val}) ---")
print(f"Número total de vértices (n) = {n}")
print(f"Número de vértices internos (i) = {i}")
print(f"Número de hojas (l) = {l}")
print(f"Número de aristas (n - 1) = {n - 1}")
print(f"Altura mínima requerida h >= ceil(log_m(l)) = {h_min}")

# Verificación de igualdad n = m*i + 1
print(f"¿Se cumple n == m*i + 1? {n == m * i + 1}")

```

## 3. Árboles Binarios de Búsqueda (ABB / BST)

Un **Árbol Binario de Búsqueda (ABB)** es un árbol binario ordenado cuyas claves cumplen la propiedad fundamental de ordenamiento:
> Para cualquier nodo $N$, todas las claves del subárbol izquierdo de $N$ son **menores** que la clave de $N$ ($Clave(Izquierda) < Clave(N)$), y todas las claves del subárbol derecho de $N$ son **mayores** ($Clave(Derecha) > Clave(N)$).

### 3.1 Operaciones Principales
1. **Búsqueda:** Compara la clave buscada con la raíz. Si es igual, concluye con éxito. Si es menor, busca en el subárbol izquierdo; si es mayor, busca en el derecho. Complejidad promedio $O(\log N)$.
2. **Inserción:** Realiza una búsqueda infructuosa hasta encontrar la posición libre (hoja vacía) que mantenga la propiedad de ordenamiento e inserta el nuevo nodo.
3. **Borrado:** Operación compleja que requiere reestructuración. Se dividen en tres escenarios:
   - **Caso 1: Nodo sin hijos (Hoja):** Se desconecta directamente de su padre.
   - **Caso 2: Nodo con un solo hijo:** Se elimina el nodo y su hijo se conecta directamente al padre del nodo eliminado (el hijo toma su lugar).
   - **Caso 3: Nodo con dos hijos:** Se reemplaza la clave del nodo por la de su **predecesor en inorden** (el valor máximo en su subárbol izquierdo) o su **sucesor en inorden** (el valor mínimo en su subárbol derecho). Posteriormente se elimina de forma recursiva dicho predecesor/sucesor (que caerá en el caso 1 o 2).

### 3.2 Aclaración Académica de la Gráfica del PDF
En la **Figura 18** (página 9) del PDF oficial, se ilustra la operación "Borrar un nodo con un subárbol hijo". La gráfica muestra la eliminación del nodo **$70$** (cuyo único hijo es $74$). Tras la eliminación, el nodo resultante debería pasar a ser el **$74$** colgado del nodo $59$. Sin embargo, en el dibujo de salida se observa un error tipográfico en la etiqueta de la burbuja, donde figura un **$70$** en lugar de **$74$**.
Nuestra simulación en Python corregirá algebraicamente este error, demostrando que la estructura correcta reemplaza al nodo $70$ por su hijo $74$.


```python
class BSTNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None
        
    def insert(self, key):
        self.root = self._insert_rec(self.root, key)
        
    def _insert_rec(self, root, key):
        if root is None:
            return BSTNode(key)
        if key < root.key:
            root.left = self._insert_rec(root.left, key)
        elif key > root.key:
            root.right = self._insert_rec(root.right, key)
        return root
        
    def search(self, key):
        return self._search_rec(self.root, key)
        
    def _search_rec(self, root, key):
        if root is None or root.key == key:
            return root
        if key < root.key:
            return self._search_rec(root.left, key)
        return self._search_rec(root.right, key)
        
    def delete(self, key):
        self.root = self._delete_rec(self.root, key)
        
    def _delete_rec(self, root, key):
        if root is None:
            return root
            
        if key < root.key:
            root.left = self._delete_rec(root.left, key)
        elif key > root.key:
            root.right = self._delete_rec(root.right, key)
        else:
            # Caso 1 y 2: Un solo hijo o sin hijos
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
                
            # Caso 3: Dos hijos (Sucesor en inorden: mínimo del subárbol derecho)
            temp = self._min_value_node(root.right)
            root.key = temp.key
            root.right = self._delete_rec(root.right, temp.key)
            
        return root
        
    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    def inorder(self):
        res = []
        self._inorder_rec(self.root, res)
        return res
        
    def _inorder_rec(self, root, res):
        if root:
            self._inorder_rec(root.left, res)
            res.append(root.key)
            self._inorder_rec(root.right, res)

    def print_bst(self, node=None, prefix="", is_left=True):
        if node is None:
            node = self.root
            if node is None:
                print("Árbol vacío")
                return
        if node.right:
            self.print_bst(node.right, prefix + (" |   " if is_left else "    "), False)
        print(prefix + (" +-- " if is_left else " +-- ") + str(node.key))
        if node.left:
            self.print_bst(node.left, prefix + ("    " if is_left else " |   "), True)

# 1. Simulación de la Inserción Secuencial (Ejemplo de la pág. 8)
# Claves a insertar: 4, 16, 3, 10, 20, 7, 12, 5
claves = [4, 16, 3, 10, 20, 7, 12, 5]
tree = BinarySearchTree()
for k in claves:
    tree.insert(k)

print("--- Estructura del ABB después de insertar [4, 16, 3, 10, 20, 7, 12, 5] ---")
tree.print_bst()
print(f"Recorrido inorden (debe estar ordenado): {tree.inorder()}")

# 2. Simulación de Búsqueda
target = 10
found = tree.search(target)
print(f"\nBúsqueda de la clave {target}: {'Encontrada' if found else 'No encontrada'}")

# 3. Simulación de Borrado (Caso 1: Hoja - Borrar 5)
print("\n--- Borrado de clave 5 (Nodo Hoja) ---")
tree.delete(5)
tree.print_bst()

# 4. Simulación de Borrado (Caso 2: Un hijo - Borrar 12)
print("\n--- Borrado de clave 12 (Nodo con un hijo, su lugar lo toma el 10? No, 12 es hijo de 10) ---")
# En el árbol actual: 10 tiene como hijos 7 (izq) y 12 (der). Si borramos 12 (hoja), 10 queda con un hijo 7.
# Borremos 10 (que tiene hijo 7). Su hijo 7 pasará a colgar de 16 directamente
tree.delete(10)
tree.print_bst()

```

## 4. Árboles de Decisión y Complejidad Asintótica

Un **árbol de decisión** es un árbol con raíz en el que cada nodo interno representa una comparación o pregunta binaria (por ejemplo, $a \le b$), y las hojas representan los resultados posibles del procedimiento.

### 4.1 Cota Inferior para Algoritmos de Ordenación
Cualquier algoritmo que ordene una lista de $n$ elementos mediante comparaciones de elementos dos a dos se puede representar mediante un árbol de decisión binario:
* **Hojas:** Cada hoja representa una de las permutaciones de ordenamiento posibles de los $n$ datos. Dado que existen $n!$ permutaciones posibles, el árbol de decisión debe tener al menos $n!$ hojas:
  $$l \ge n!$$
* **Altura:** Por propiedades métricas del árbol binario, la altura $h$ (número de comparaciones en el peor caso) cumple con:
  $$h \ge \lceil \log_2 l \rceil \implies h \ge \lceil \log_2(n!) \rceil$$

### 4.2 Aproximación asintótica (Fórmula de Stirling)
Aplicando aproximaciones matemáticas de logaritmos de factoriales:
$$\log_2(n!) = \sum_{k=1}^n \log_2 k \ge \sum_{k=\lceil n/2 \rceil}^n \log_2 k \ge \frac{n}{2} \log_2 \left( \frac{n}{2} \right) = \frac{n}{2} (\log_2 n - 1) = \Omega(n \log n)$$

* **Teorema:** El número de comparaciones promedio y en el peor caso de cualquier algoritmo de ordenación por comparación binaria es **$\Omega(n \log n)$**. Esto demuestra matemáticamente que algoritmos como MergeSort y QuickSort son asintóticamente óptimos.


## 5. Algoritmos de Recorrido de Árboles

Para acceder de manera sistemática y visitar cada uno de los vértices de un árbol ordenado con raíz, se definen tres algoritmos recursivos clásicos:

### 5.1 Recorrido en Preorden (Preorder)
1. Visitar la raíz $r$.
2. Recorrer el primer subárbol $T_1$ en preorden.
3. Recorrer el segundo subárbol $T_2$ en preorden (y así sucesivamente de izquierda a derecha).
> **Esquema:** Raíz $\to$ Izquierda $\to$ Derecha.

### 5.2 Recorrido en Inorden (Inorder)
1. Recorrer el primer subárbol $T_1$ en inorden.
2. Visitar la raíz $r$.
3. Recorrer el resto de subárboles $T_2, \dots, T_n$ en inorden.
> **Esquema:** Izquierda $\to$ Raíz $\to$ Derecha.

### 5.3 Recorrido en Postorden (Postorder)
1. Recorrer el primer subárbol $T_1$ en postorden.
2. Recorrer el segundo subárbol $T_2$ en postorden (y así sucesivamente de izquierda a derecha).
3. Visitar la raíz $r$.
> **Esquema:** Izquierda $\to$ Derecha $\to$ Raíz.

### 5.4 Simulación del Árbol de la Figura 21
El árbol de la **Figura 21** (página 11) tiene las siguientes conexiones:
* Raíz: $2_a$ (denotado $2_a$ para distinguirlo del nodo hoja $2_b$).
* Hijos de $2_a$: $7$ y $5_a$.
* Hijos de $7$: $2_b$ y $6$.
* Hijos de $6$: $5_b$ y $11$.
* Hijo de $5_a$: $9$.
* Hijo de $9$: $4$.

A continuación programaremos los algoritmos recursivos y comprobaremos que las secuencias de nodos resultantes coinciden al 100% con las listas provistas en el PDF.


```python
class TraversalNode:
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None

# Definir la estructura exacta del árbol binario de la Figura 21
# Nombres de los nodos tal como se listan en las salidas
# Para que los strings de comparación coincidan con los números, usamos enteros.
# Preorden esperado: [2, 7, 2, 6, 5, 11, 5, 9, 4]
# Inorden esperado: [2, 7, 5, 6, 11, 2, 5, 4, 9]
# Postorden esperado: [2, 5, 11, 6, 7, 4, 9, 5, 2]

node_2a = TraversalNode(2)
node_7 = TraversalNode(7)
node_5a = TraversalNode(5)
node_2b = TraversalNode(2)
node_6 = TraversalNode(6)
node_5b = TraversalNode(5)
node_11 = TraversalNode(11)
node_9 = TraversalNode(9)
node_4 = TraversalNode(4)

# Enlaces binarios (Hijo izquierdo y derecho)
node_2a.left = node_7
node_2a.right = node_5a

node_7.left = node_2b
node_7.right = node_6

node_6.left = node_5b
node_6.right = node_11

node_5a.right = node_9
node_9.left = node_4

def preorder(node):
    if node is None:
        return []
    return [node.name] + preorder(node.left) + preorder(node.right)

def inorder(node):
    if node is None:
        return []
    return inorder(node.left) + [node.name] + inorder(node.right)

def postorder(node):
    if node is None:
        return []
    return postorder(node.left) + postorder(node.right) + [node.name]

pre_res = preorder(node_2a)
in_res = inorder(node_2a)
post_res = postorder(node_2a)

print("--- Validación de Recorridos en Árboles (Figura 21) ---")
print(f"Preorden obtenido: {pre_res}")
print(f"Preorden esperado: [2, 7, 2, 6, 5, 11, 5, 9, 4]")
print(f"¿Coincide Preorden? {pre_res == [2, 7, 2, 6, 5, 11, 5, 9, 4]}")
print("-" * 60)
print(f"Inorden obtenido:  {in_res}")
print(f"Inorden esperado:  [2, 7, 5, 6, 11, 2, 5, 4, 9]")
print(f"¿Coincide Inorden?  {in_res == [2, 7, 5, 6, 11, 2, 5, 4, 9]}")
print("-" * 60)
print(f"Postorden obtenido: {post_res}")
print(f"Postorden esperado: [2, 5, 11, 6, 7, 4, 9, 5, 2]")
print(f"¿Coincide Postorden? {post_res == [2, 5, 11, 6, 7, 4, 9, 5, 2]}")

```

## 6. Resumen de la Lección

1. **Estructura Jerárquica:** Un árbol es un grafo no dirigido simple, conexo y acíclico. Múltiples árboles disjuntos forman un bosque.
2. **Propiedades Estructurales:** Los árboles poseen métricas fundamentales como altura, peso, nivel, orden y grado. Un teorema clave establece que un árbol de $n$ vértices contiene exactamente $n-1$ aristas.
3. **Árboles m-arios Completos:** Tienen relaciones cuantitativas precisas entre el número total de vértices ($n$), los vértices internos ($i$), y el número de hojas ($l$).
4. **Árboles Binarios de Búsqueda (ABB):** Mantienen una relación de orden estricta que permite buscar elementos en tiempo promedio logarítmico. Su borrado requiere tres casos de reestructuración (incluyendo el reemplazo por predecesor/sucesor).
5. **Árboles de Decisión y Complejidad:** El modelado de algoritmos de comparación binaria mediante árboles demuestra rigurosamente que ordenar $n$ elementos tiene una cota inferior asintótica infranqueable de $\Omega(n \log n)$.
6. **Recorridos Recursivos:** La exploración ordenada se implementa sistemáticamente mediante los esquemas de Preorden, Inorden y Postorden, los cuales se validaron de forma exacta en la estructura provista.

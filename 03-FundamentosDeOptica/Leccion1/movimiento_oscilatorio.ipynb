## 1. Movimientos Periódicos y Repaso Trigonométrico

Un **movimiento periódico** es aquel que se repite en intervalos regulares de tiempo, denominados **periodo** ($T$). Históricamente, el ser humano ha observado fenómenos periódicos en el entorno natural: las fases de la luna, las estaciones climáticas, las mareas oceánicas y las crecidas de los ríos. ## 2. Movimiento Armónico Simple (MAS)

El **Movimiento Armónico Simple (MAS)** es un modelo físico para un oscilador libre y sin pérdidas donde la aceleración (o fuerza neta) es proporcional al desplazamiento respecto a la posición de equilibrio y actúa en dirección opuesta a este:
$$a = -c x$$
Donde $c$ es una constante positiva del sistema.

### 2.1 Ecuación de un Resorte (Ley de Hooke)
Para un bloque de masa $m$ sujeto a un resorte ideal de constante de elasticidad $k$, la fuerza restauradora sigue la **Ley de Hooke**:
$$F = -kx$$
Aplicando la segunda ley de Newton:
$$F = m a \implies m rac{d^2x}{dt^2} = -kx \implies rac{d^2x}{dt^2} + rac{k}{m}x = 0$$

Definiendo la **frecuencia angular natural** $\omega = \sqrt{rac{k}{m}}$, la ecuación diferencial toma la forma estándar:
$$rac{d^2x}{dt^2} + \omega^2 x = 0$$

La solución analítica de esta ecuación diferencial es de tipo armónico:
$$x(t) = A \cos(\omega t + \Phi)$$
Donde:
* $A$ es la **Amplitud** (máximo desplazamiento desde el equilibrio).
* $\omega = 2\pi f = rac{2\pi}{T}$ es la frecuencia angular (rad/s).
* $\Phi$ es la constante de fase determinada por las condiciones iniciales de posición y velocidad en $t=0$.

A partir de la posición $x(t)$, se obtienen las expresiones para la velocidad $v(t)$ y la aceleración $a(t)$ por diferenciación sucesiva respecto al tiempo:
* **Velocidad ($v$):**
  $$v(t) = rac{dx}{dt} = -A \omega \sin(\omega t + \Phi)$$
* **Aceleración ($a$):**
  $$a(t) = rac{dv}{dt} = -A \omega^2 \cos(\omega t + \Phi) = -\omega^2 x(t)$$

### 2.2 Conservación de la Energía Mecánica
En un sistema libre de fricción y pérdidas, la energía mecánica total ($E_T$) se conserva constante a lo largo del tiempo, transformándose continuamente entre energía cinética y potencial:
* **Energía Cinética ($E_c$):** Depende de la velocidad lineal:
  $$E_c(t) = rac{1}{2} m v(t)^2 = rac{1}{2} m A^2 \omega^2 \sin^2(\omega t + \Phi)$$
* **Energía Potencial ($E_p$):** Para el resorte elástico, depende de la elongación:
  $$E_p(t) = rac{1}{2} k x(t)^2 = rac{1}{2} k A^2 \cos^2(\omega t + \Phi)$$

Como $\omega^2 = k/m \implies m\omega^2 = k$, podemos simplificar la suma total:
$$E_T = E_c(t) + E_p(t) = rac{1}{2} k A^2 \sin^2(\omega t + \Phi) + rac{1}{2} k A^2 \cos^2(\omega t + \Phi)$$
$$E_T = rac{1}{2} k A^2 \left[ \sin^2(\omega t + \Phi) + \cos^2(\omega t + \Phi) ight] = rac{1}{2} k A^2$$

Dado que $k$ y $A$ son constantes, la energía total permanece invariable independientemente de la posición y velocidad instantánea del sistema.


```python
# Parámetros del sistema físico
m = 1.5      # Masa (kg)
k = 24.0     # Constante del resorte (N/m)
A = 0.4      # Amplitud (m)
phi = np.pi/3 # Constante de fase (rad)

omega = np.sqrt(k / m)
T = 2 * np.pi / omega
t = np.linspace(0, 3 * T, 1000)

# Ecuaciones cinemáticas
x = A * np.cos(omega * t + phi)
v = -A * omega * np.sin(omega * t + phi)
a = -A * (omega**2) * np.cos(omega * t + phi)

# Ecuaciones de energía
Ec = 0.5 * m * (v**2)
Ep = 0.5 * k * (x**2)
Et = Ec + Ep

# Generación del gráfico con subplots (Panel de 2 subplots verticales)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plot 1: Cinemática (Posición y Velocidad)
ax1.plot(t, x, 'r-', linewidth=2.5, label='Posición $x(t)$ (m)')
ax1.plot(t, v, 'b--', linewidth=2, label='Velocidad $v(t)$ (m/s)')
ax1.set_title('Cinemática del Movimiento Armónico Simple', fontsize=13, fontweight='bold')
ax1.set_ylabel('Amplitud Cinemática', fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(frameon=True, loc='upper right')

# Plot 2: Energías (Cinética, Potencial, Total)
ax2.plot(t, Ec, 'g-', linewidth=2, label='Energía Cinética $E_c$ (J)')
ax2.plot(t, Ep, 'orange', linewidth=2, label='Energía Potencial $E_p$ (J)')
ax2.plot(t, Et, 'black', linewidth=2.5, linestyle=':', label='Energía Total $E_T$ (J)')
ax2.set_title('Conservación de la Energía Mecánica', fontsize=13, fontweight='bold')
ax2.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax2.set_ylabel('Energía (Joules)', fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(frameon=True, loc='center right')

plt.tight_layout()
plt.show()
```

## 3. Estudio del Péndulo Simple

Un **péndulo simple** consiste en una masa puntual $m$ suspendida de un pivote rígido por medio de una cuerda de longitud $l$, inextensible y de masa despreciable. Al desplazar la masa de su punto de equilibrio y soltarla, la gravedad actúa como fuerza restauradora, provocando la oscilación.

### 3.1 Ecuaciones de Movimiento
Descomponiendo las fuerzas en la dirección tangencial al arco de trayectoria descrita ($s = l 	heta$), la segunda ley de Newton dicta:
$$F_t = -mg \sin	heta = m a_t$$

Donde $a_t = rac{d^2s}{dt^2} = l rac{d^2	heta}{dt^2}$. Reemplazando esta expresión, se obtiene:
$$-mg \sin	heta = m l rac{d^2	heta}{dt^2} \implies rac{d^2	heta}{dt^2} + rac{g}{l} \sin	heta = 0$$

Esta es la ecuación diferencial exacta que describe el péndulo simple libre. Debido al término no lineal $\sin	heta$, no existe una solución en términos de funciones elementales.

### 3.2 Aproximación del Ángulo Pequeño
Para ángulos pequeños ($	heta$ en radianes, generalmente $	heta < 15^\circ$), podemos aproximar el seno mediante su expansión de Taylor de primer orden:
$$\sin	heta pprox 	heta$$

Aplicando esta linealización, la ecuación se simplifica a la forma del MAS:
$$rac{d^2	heta}{dt^2} + rac{g}{l} 	heta = 0$$

Definiendo la frecuencia angular $\omega = \sqrt{rac{g}{l}}$, la solución general para una velocidad angular inicial nula es:
$$	heta(t) = 	heta_0 \cos(\omega t)$$

El **periodo del péndulo simple** es, por tanto:
$$P = rac{2\pi}{\omega} = 2\pi \sqrt{rac{l}{g}}$$

Esta ecuación demuestra el principio de **isocronismo**: para amplitudes pequeñas, el periodo es independiente del ángulo inicial $	heta_0$ y de la masa $m$.


```python
from scipy.integrate import solve_ivp

# Parámetros físicos
g = 9.81  # Aceleración de la gravedad (m/s^2)
l = 1.0   # Longitud de la cuerda (m)
omega_lin = np.sqrt(g / l)

# Definición del sistema de ecuaciones diferenciales exactas (no lineales)
def pendulo_exacto(t, y):
    theta, omega = y
    return [omega, -(g / l) * np.sin(theta)]

# Tiempo de simulación (3 periodos teóricos)
T_teorico = 2 * np.pi / omega_lin
t_span = (0, 3 * T_teorico)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Simulación para dos ángulos iniciales diferentes (Pequeño y Grande)
theta_0_peq = np.radians(10.0)  # 10 grados
theta_0_gra = np.radians(60.0)  # 60 grados

sol_peq = solve_ivp(pendulo_exacto, t_span, [theta_0_peq, 0], t_eval=t_eval)
sol_gra = solve_ivp(pendulo_exacto, t_span, [theta_0_gra, 0], t_eval=t_eval)

# Soluciones analíticas lineales (MAS)
theta_lin_peq = theta_0_peq * np.cos(omega_lin * t_eval)
theta_lin_gra = theta_0_gra * np.cos(omega_lin * t_eval)

# Graficación comparativa
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Subplot 1: Ángulo Pequeño (10°)
ax1.plot(sol_peq.t, np.degrees(sol_peq.y[0]), 'b-', linewidth=2.5, label='Exacto (Numérico)')
ax1.plot(t_eval, np.degrees(theta_lin_peq), 'r--', linewidth=2, label='Lineal ($\\sin\\theta \\approx \\theta$)')
ax1.set_title('Péndulo con Ángulo Pequeño ($\\theta_0 = 10^\\circ$)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax1.set_ylabel('Ángulo $\\theta$ (grados)', fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(frameon=True)

# Subplot 2: Ángulo Grande (60°)
ax2.plot(sol_gra.t, np.degrees(sol_gra.y[0]), 'b-', linewidth=2.5, label='Exacto (Numérico)')
ax2.plot(t_eval, np.degrees(theta_lin_gra), 'r--', linewidth=2, label='Lineal ($\\sin\\theta \\approx \\theta$)')
ax2.set_title('Péndulo con Ángulo Grande ($\\theta_0 = 60^\\circ$)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax2.set_ylabel('Ángulo $\\theta$ (grados)', fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

## 4. Estudio del Péndulo Físico

El **péndulo físico** (o péndulo compuesto) es un cuerpo rígido de forma arbitraria con masa total $m$, libre de rotar en un plano vertical alrededor de un eje horizontal fijo que pasa por un pivote. A diferencia del péndulo simple, la distribución de la masa a lo largo de su geometría influye en su inercia rotacional.

### 4.1 Torque y Momento de Inercia
Definimos:
* $l$: Distancia entre el pivote de suspensión y el centro de masa (CM) del cuerpo rígido.
* $I$: Momento de inercia rotacional del cuerpo rígido respecto al eje de rotación que pasa por el pivote.

El torque $\tau$ ejercido por la gravedad (concentrado en el CM del cuerpo) con respecto al pivote es:
$$\tau = -mgl \sin	heta$$

De la mecánica rotacional sabemos que $\tau = I \alpha$, donde $\alpha = \frac{d^2	heta}{dt^2}$ es la aceleración angular. Igualando los términos:
$$I \frac{d^2	heta}{dt^2} = -mgl \sin	heta \implies \frac{d^2	heta}{dt^2} + rac{mgl}{I} \sin	heta = 0$$

### 4.2 Solución Linealizada
Utilizando la aproximación para ángulos pequeños ($\sin	heta \approx \theta$):
$$\frac{d^2	heta}{dt^2} + \frac{mgl}{I} \theta = 0$$

Esta es la ecuación diferencial del MAS con una frecuencia angular:
$$\omega = \sqrt{\frac{mgl}{I}}$$

El **periodo del péndulo físico** es:
$$P = 2\pi \sqrt{\frac{I}{mgl}}$$

### 4.3 Reducción a Péndulo Simple
Si concentramos toda la masa en el centro de masas (masa puntual), el momento de inercia respecto al pivote se reduce a $I = ml^2$. Sustituyendo esto en la fórmula general del periodo:
$$P = 2\pi \sqrt{\frac{ml^2}{mgl}} = 2\pi \sqrt{\frac{l}{g}}$$

Esto demuestra que el péndulo simple es un caso particular del péndulo físico.


```python
# Parámetros físicos de una barra uniforme
m_barra = 2.0  # Masa de la barra (kg)
L_barra = 1.2  # Longitud total de la barra (m)
g = 9.81

# Para una barra uniforme pivotada en su extremo:
# Momento de inercia I = (1/3) * m * L^2
I_barra = (1.0 / 3.0) * m_barra * (L_barra**2)
# Distancia del pivote al CM
d_cm = L_barra / 2.0

omega_fisico = np.sqrt((m_barra * g * d_cm) / I_barra)
T_fisico = 2 * np.pi / omega_fisico

# Péndulo simple equivalente de longitud L_barra
omega_simple = np.sqrt(g / L_barra)
T_simple = 2 * np.pi / omega_simple

t_eval = np.linspace(0, 3 * T_simple, 1000)
theta_0 = np.radians(10.0)

# Soluciones analíticas
theta_fisico = theta_0 * np.cos(omega_fisico * t_eval)
theta_simple = theta_0 * np.cos(omega_simple * t_eval)

# Graficación
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(t_eval, np.degrees(theta_fisico), 'r-', linewidth=2.5, label='Péndulo Físico (Barra, $I=\\frac{1}{3}mL^2$)')
ax.plot(t_eval, np.degrees(theta_simple), 'b--', linewidth=2, label='Péndulo Simple (Longitud $L$)')
ax.set_title('Comparación de Frecuencia: Péndulo Físico vs. Péndulo Simple', fontsize=13, fontweight='bold')
ax.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax.set_ylabel('Ángulo $\\theta$ (grados)', fontsize=11)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(frameon=True)

plt.tight_layout()
plt.show()
```

## 5. Oscilaciones Amortiguadas

En los sistemas físicos reales siempre existen pérdidas de energía mecánica (rozamiento por fricción seca, resistencia viscosa en un medio fluido o pérdidas internas). Estas pérdidas se oponen al movimiento y amortiguan las oscilaciones hasta detener el sistema por completo.

### 5.1 Ecuación de Movimiento con Amortiguación Viscosa
Consideramos un amortiguador donde la fuerza disipativa es proporcional a la velocidad lineal del objeto para bajas velocidades:
$$F_d = -b v = -b rac{dx}{dt}$$
Donde $b$ es la constante de resistencia del medio o amortiguación (N·s/m).

Aplicando la ley de fuerzas al sistema masa-resorte:
$$\Sigma F = -kx - b rac{dx}{dt} = m rac{d^2x}{dt^2} \implies m rac{d^2x}{dt^2} + b rac{dx}{dt} + kx = 0$$

Dividiendo por la masa $m$, la ecuación se escribe:
$$\frac{d^2x}{dt^2} + 2\gamma \frac{dx}{dt} + \omega_0^2 x = 0$$

Donde:
* $\omega_0 = \sqrt{\frac{k}{m}}$ es la frecuencia natural no amortiguada del sistema.
* $\gamma = \frac{b}{2m}$ es el parámetro de atenuación.

### 5.2 Tipos de Amortiguamiento
Dependiendo del valor del coeficiente de amortiguamiento $b$ en comparación con la cantidad crítica $\sqrt{4mk}$, la solución matemática adopta tres comportamientos físicos distintos:

1. **Subamortiguamiento ($b < \sqrt{4mk}$):**
   El sistema oscila de forma periódica amortiguada, con una amplitud que decrece exponencialmente. La solución exacta es:
   $$x(t) = A_0 e^{-\gamma t} \cos(\omega t + \phi)$$
   Donde la frecuencia angular amortiguada es menor que la natural:
   $$\omega = \sqrt{\omega_0^2 - \gamma^2} = \sqrt{\frac{k}{m} - \left(\frac{b}{2m}\right)^2}$$

2. **Amortiguamiento Crítico ($b = \sqrt{4mk}$):**
   El sistema no realiza ninguna oscilación completa; regresa al punto de equilibrio en el menor tiempo posible sin sobrepasarlo. La solución general es:
   $$x(t) = (C_1 + C_2 t) e^{-\gamma t}$$

3. **Sobreamortiguamiento ($b > \sqrt{4mk}$):**
   El sistema es sumamente viscoso, retornando lentamente al equilibrio de forma asintótica y sin oscilar. La solución general está compuesta por dos exponenciales puramente reales y negativas:
   $$x(t) = C_1 e^{r_1 t} + C_2 e^{r_2 t}$$
   Donde:
   $$r_{1,2} = -\gamma \pm \sqrt{\gamma^2 - \omega_0^2}$$


```python
# Parámetros base
m = 1.0       # Masa (kg)
k = 25.0      # Constante del resorte (N/m)
omega_0 = np.sqrt(k / m)  # 5.0 rad/s

# Amortiguamiento crítico teórico
b_crit = np.sqrt(4 * m * k)  # sqrt(100) = 10.0 N·s/m

t = np.linspace(0, 5, 1000)

# Condiciones iniciales: x(0) = 1.0 m, v(0) = 0.0 m/s
x0 = 1.0
v0 = 0.0

# 1. Regímen Subamortiguado (b = 1.5 < 10.0)
b_sub = 1.5
gamma_sub = b_sub / (2 * m)
omega_sub = np.sqrt(omega_0**2 - gamma_sub**2)
phi_sub = np.arctan(-gamma_sub / omega_sub)
A_sub = x0 / np.cos(phi_sub)
x_sub = A_sub * np.exp(-gamma_sub * t) * np.cos(omega_sub * t + phi_sub)

# 2. Regímen Amortiguado Crítico (b = b_crit = 10.0)
gamma_crit = b_crit / (2 * m)
C1_crit = x0
C2_crit = v0 + gamma_crit * x0
x_crit = (C1_crit + C2_crit * t) * np.exp(-gamma_crit * t)

# 3. Regímen Sobreamortiguado (b = 16.0 > 10.0)
b_sob = 16.0
gamma_sob = b_sob / (2 * m)
r1 = -gamma_sob + np.sqrt(gamma_sob**2 - omega_0**2)
r2 = -gamma_sob - np.sqrt(gamma_sob**2 - omega_0**2)
C2_sob = (v0 - r1 * x0) / (r2 - r1)
C1_sob = x0 - C2_sob
x_sob = C1_sob * np.exp(r1 * t) + C2_sob * np.exp(r2 * t)

# Graficación
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, x_sub, 'b-', linewidth=2.5, label='Subamortiguado ($b = 1.5$)')
ax.plot(t, x_crit, 'g-', linewidth=2.5, label='Amortiguamiento Crítico ($b = 10.0$)')
ax.plot(t, x_sob, 'r-', linewidth=2.5, label='Sobreamortiguado ($b = 16.0$)')
ax.plot(t, x0 * np.exp(-gamma_sub * t), 'k--', alpha=0.5, label='Envolvente exponencial Sub')
ax.plot(t, -x0 * np.exp(-gamma_sub * t), 'k--', alpha=0.5)

ax.set_title('Evolución Temporal de los Tres Regímenes de Amortiguamiento', fontsize=13, fontweight='bold')
ax.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax.set_ylabel('Desplazamiento $x$ (m)', fontsize=11)
ax.axhline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(frameon=True)

plt.tight_layout()
plt.show()
```

## 6. Oscilaciones Forzadas y Resonancia

Si un oscilador amortiguado se somete a una fuerza externa periódica oscilante (fuerza impulsora), el sistema experimentará **oscilaciones forzadas**. La fuerza externa de excitación armónica se escribe:
$$F_{ext}(t) = F_0 \sin(\omega t)$$
Donde $\omega$ es la frecuencia angular impulsora impuesta externamente.

### 6.1 Ecuación de Movimiento Forzado
Sumando las fuerzas en la masa:
$$m rac{d^2x}{dt^2} + b rac{dx}{dt} + kx = F_0 \sin(\omega t)$$

Esta es una ecuación diferencial no homogénea. La solución general está compuesta por una parte transitoria (la solución homogénea que decae exponencialmente a cero debido a la amortiguación) y una solución particular en **estado estacionario** donde el sistema oscila de forma periódica con la misma frecuencia $\omega$ de la fuerza impulsora:
$$x(t) = A \cos(\omega t + \phi)$$

Sustituyendo esta solución en la ecuación diferencial y resolviendo mediante identidades trigonométricas o números complejos, se deduce la amplitud $A$ en estado estacionario:
$$A(\omega) = rac{F_0}{\sqrt{b^2 \omega^2 + m^2(\omega^2 - \omega_0^2)^2}}$$

Dividiendo por $m$:
$$A(\omega) = rac{F_0/m}{\sqrt{4\gamma^2\omega^2 + (\omega^2 - \omega_0^2)^2}}$$

### 6.2 Resonancia de Amplitud
La **resonancia** ocurre cuando la amplitud de la respuesta $A(\omega)$ alcanza su valor máximo. Para evaluar el pico de la curva, analizamos el denominador. En ausencia de amortiguación ($b = 0$), si la frecuencia impulsora iguala la frecuencia natural del sistema ($\omega = \omega_0$), la amplitud teórica tiende a infinito:
$$A(\omega_0) = rac{F_0}{b\omega_0}$$

En sistemas reales, la amplitud es acotada por $b$, pero si la fricción es muy pequeña, la amplitud incrementa dramáticamente en la frecuencia de resonancia. Este fenómeno es crucial en ingeniería estructural (evitar el colapso de puentes como el de Tacoma Narrows en 1940) y óptica (absorción de luz por resonancias electrónicas y atómicas en materiales).


```python
# Parámetros del sistema
m = 1.0
k = 100.0
omega_0 = np.sqrt(k / m)  # 10.0 rad/s
F_0 = 10.0

# Rango de frecuencias impulsoras normalizadas (omega / omega_0)
omega_ratios = np.linspace(0.1, 2.0, 1000)
omegas = omega_ratios * omega_0

# Coeficientes de amortiguación para simular
damping_coefficients = [0.5, 1.0, 2.0, 5.0, 10.0]

fig, ax = plt.subplots(figsize=(10, 6))

for b in damping_coefficients:
    # Ecuación analítica de la amplitud en estado estacionario
    A = F_0 / np.sqrt((b**2) * (omegas**2) + (m**2) * ((omegas**2 - omega_0**2)**2))
    ax.plot(omega_ratios, A, linewidth=2, label=f'$b = {b}$ N·s/m')

ax.set_title('Respuesta de Amplitud de un Oscilador Forzado (Resonancia)', fontsize=13, fontweight='bold')
ax.set_xlabel('Relación de Frecuencia $\\omega / \\omega_0$', fontsize=11)
ax.set_ylabel('Amplitud del Sistema $A$ (m)', fontsize=11)
ax.axvline(1.0, color='gray', linestyle=':', label='Resonancia teórica $\\omega = \\omega_0$')
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(frameon=True, fontsize=10)

plt.tight_layout()
plt.show()
```

## 1. Descripción del Movimiento Ondulatorio y la Ecuación de Onda

Una **onda** es una perturbación que viaja a través de un espacio o medio, transfiriendo energía de un punto a otro. Dependiendo de la naturaleza del medio, clasificamos las ondas en:
* **Mecánicas:** Requieren un medio material elástico para propagarse (ej. el sonido en el aire, ondas elásticas en cuerdas).
* **Electromagnéticas:** No requieren un medio físico; consisten en oscilaciones acopladas de campos eléctricos y magnéticos que se propagan en el vacío a la velocidad de la luz $c$.
* **Gravitacionales:** Ondas asociadas con la curvatura del espacio-tiempo producidas por aceleraciones de masas masivas (ej. colisiones de agujeros negros).

### 1.1 Parámetros de una Onda
Para caracterizar matemáticamente una onda periódica en el plano cartesiano $x(t)$, definimos las siguientes variables:
* **Elongación ($u$):** Posición instantánea de la partícula respecto al equilibrio.
* **Cresta y Valle:** Puntos de máxima y mínima elongación, respectivamente.
* **Nodo:** Puntos donde la elongación es nula ($u = 0$).
* **Amplitud ($A$):** Valor de la elongación máxima.
* **Longitud de onda ($\lambda$):** Distancia física mínima entre dos puntos idénticos en fase (como dos crestas sucesivas).
* **Periodo ($T$) y Frecuencia ($f$):** Tiempo para una oscilación completa y número de oscilaciones por unidad de tiempo ($f = 1/T$, medido en Hertz).
* **Número de onda ($k$):** Medida de la frecuencia espacial: $k = \frac{2\pi}{\lambda}$.
* **Frecuencia angular ($\omega$):** Velocidad angular de fase: $\omega = 2\pi f = \frac{2\pi}{T}$.
* **Velocidad de propagación ($v$):** Velocidad a la que se desplaza la perturbación:
  $$v = \frac{\lambda}{T} = \frac{\omega}{k}$$

### 1.2 Derivación de la Ecuación de Onda Unidimensional
Para derivar la ecuación diferencial de onda, imaginamos una cadena lineal de $N$ masas $m$ interconectadas por resortes idénticos de longitud en reposo $h$ y constante elástica $k_s$.

Sea $u(x, t)$ el desplazamiento de la masa en la posición $x$. La masa vecina a la derecha se sitúa en $x+h$ con desplazamiento $u(x+h, t)$, y la de la izquierda en $x-h$ con desplazamiento $u(x-h, t)$.

Aplicando la ley de Hooke para las fuerzas netas sobre la masa en la posición $x+h$:
$$F = k_s [u(x+2h, t) - u(x+h, t)] - k_s [u(x+h, t) - u(x, t)]$$

Por la segunda ley de Newton ($F = m a$):
$$m \frac{\partial^2 u(x+h, t)}{\partial t^2} = k_s [u(x+2h, t) - 2u(x+h, t) + u(x, t)]$$

Considerando todo el sistema distribuido a lo largo de una longitud física total $L$ y masa total $M$, donde $L = Nh$ y $M = Nm$:
* Constante elástica macroscópica: $K = k_s / N \implies k_s = K N = K \frac{L}{h}$.
* Masa de cada elemento: $m = M / N = M \frac{h}{L}$.

Reemplazando estos parámetros macroscópicos en la ecuación de movimiento:
$$\left(M \frac{h}{L}\right) \frac{\partial^2 u}{\partial t^2} = \left(K \frac{L}{h}\right) [u(x+2h, t) - 2u(x+h, t) + u(x, t)]$$
$$\frac{\partial^2 u}{\partial t^2} = \frac{K L^2}{M} \frac{u(x+2h, t) - 2u(x+h, t) + u(x, t)}{h^2}$$

Si tomamos el límite continuo donde el número de masas tiende a infinito ($N \to \infty$) y la distancia entre ellas a cero ($h \to 0$):
$$\lim_{h \to 0} \frac{u(x+2h, t) - 2u(x+h, t) + u(x, t)}{h^2} = \frac{\partial^2 u(x, t)}{\partial x^2}$$

Por lo tanto, la ecuación diferencial de onda toma la forma clásica:
$$\frac{\partial^2 u(x, t)}{\partial t^2} = v^2 \frac{\partial^2 u(x, t)}{\partial x^2}$$
Donde la velocidad de propagación $v$ se define por las propiedades elásticas e inerciales del medio:
$$v^2 = \frac{K L^2}{M} \implies v = L \sqrt{\frac{K}{M}}$$

Para una cuerda tensa con tensión $T$ y densidad lineal de masa $\mu = M/L$, la velocidad es:
$$v = \sqrt{\frac{T}{\mu}}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros espaciotemporales
L = 10.0      # Longitud de la cuerda (m)
Nx = 200      # Número de puntos espaciales
dx = L / (Nx - 1)
x = np.linspace(0, L, Nx)

v = 2.0       # Velocidad de propagación de la onda (m/s)
C = 0.5       # Número de Courant (C = v * dt / dx <= 1 para estabilidad)
dt = C * dx / v
Nt = 300      # Número de pasos de tiempo

# Inicialización de la elongación de la cuerda
u = np.zeros((Nt, Nx))

# Condición inicial: Pulso Gaussiano en el centro de la cuerda
sigma = 0.5
x0 = L / 2
u[0, :] = np.exp(-((x - x0)**2) / (2 * sigma**2))

# Segunda condición inicial (velocidad inicial nula): u_t = 0
for i in range(1, Nx - 1):
    u[1, i] = u[0, i] + 0.5 * (C**2) * (u[0, i+1] - 2 * u[0, i] + u[0, i-1])

# Resolutor por diferencias finitas (esquema explícito en el tiempo)
for n in range(1, Nt - 1):
    for i in range(1, Nx - 1):
        u[n+1, i] = 2 * u[n, i] - u[n-1, i] + (C**2) * (u[n, i+1] - 2 * u[n, i] + u[n, i-1])
    # Condiciones de contorno fijas (extremos amarrados): u(0, t) = u(L, t) = 0
    u[n+1, 0] = 0.0
    u[n+1, -1] = 0.0

# Graficación de perfiles temporales seleccionados
fig, ax = plt.subplots(figsize=(10, 6))

times_to_plot = [0, 20, 60, 100, 150, 200]
for step in times_to_plot:
    ax.plot(x, u[step, :], label=f't = {step*dt:.2f} s (Paso {step})', linewidth=1.8)

ax.set_title('Evolución de un Pulso de Onda en Cuerda con Extremos Fijos', fontsize=13, fontweight='bold')
ax.set_xlabel('Posición x (m)', fontsize=11)
ax.set_ylabel('Elongación u(x, t) (m)', fontsize=11)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(frameon=True, fontsize=9)

plt.tight_layout()
plt.show()
```

## 2. Ondas Transversales y Longitudinales

Dependiendo de la dirección de oscilación de las partículas respecto a la dirección de avance de la perturbación, clasificamos las ondas en dos categorías fundamentales:

* **Ondas Transversales:** La oscilación de las partículas es perpendicular a la dirección de propagación (ej. oscilación de una cuerda tensa, ondas electromagnéticas).
* **Ondas Longitudinales:** La oscilación de las partículas es paralela a la dirección de propagación (ej. el sonido, compresiones en un resorte helicoidal). Se transmiten a través del medio mediante variaciones de densidad y presión (compresiones y rarefacciones).

### 2.1 Polarización de las Ondas Transversales
Una propiedad exclusiva de las ondas transversales es la **polarización**, la cual describe la orientación geométrica en el espacio tridimensional de las oscilaciones. 

Para las ondas de luz, definimos el estado de polarización en base a la oscilación de su campo eléctrico $\vec{E}$:
1. **Polarización Lineal:** El vector del campo eléctrico oscila únicamente a lo largo de una línea recta contenida en un plano fijo.
2. **Polarización Circular:** El vector de campo mantiene su magnitud constante, pero su dirección rota de forma uniforme en un plano perpendicular a la propagación, describiendo una circunferencia. Puede ser levógira (giro antihorario) o dextrógira (giro horario).
3. **Polarización Elíptica:** El vector de campo eléctrico rota y altera su magnitud simultáneamente, de modo que la punta del vector dibuja una elipse en el plano de proyección.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Propagación a lo largo del eje z
z = np.linspace(0, 2 * np.pi, 200)
t = 0.0  # Captura instantánea de tiempo

# 1. Polarización Lineal (E_y en fase con E_x)
Ex_lin = np.cos(z)
Ey_lin = np.cos(z)  # En fase

# 2. Polarización Circular (Diferencia de fase = 90°)
Ex_circ = np.cos(z)
Ey_circ = np.sin(z)  # Fase desplazada pi/2

# 3. Polarización Elíptica (Amplitudes desiguales y fase desplazada)
Ex_elip = 1.5 * np.cos(z)
Ey_elip = 0.7 * np.sin(z + np.pi/4)

# Graficación 3D
fig = plt.subplots(figsize=(15, 5))

# Subplot 1: Lineal
ax1 = plt.subplot(131, projection='3d')
ax1.plot(z, Ex_lin, Ey_lin, 'r-', linewidth=2)
ax1.set_title('Polarización Lineal', fontsize=12, fontweight='bold')
ax1.set_xlabel('Eje Z (Propagación)')
ax1.set_ylabel('E_x')
ax1.set_zlabel('E_y')

# Subplot 2: Circular
ax2 = plt.subplot(132, projection='3d')
ax2.plot(z, Ex_circ, Ey_circ, 'b-', linewidth=2)
ax2.set_title('Polarización Circular', fontsize=12, fontweight='bold')
ax2.set_xlabel('Eje Z (Propagación)')
ax2.set_ylabel('E_x')
ax2.set_zlabel('E_y')

# Subplot 3: Elíptica
ax3 = plt.subplot(133, projection='3d')
ax3.plot(z, Ex_elip, Ey_elip, 'g-', linewidth=2)
ax3.set_title('Polarización Elíptica', fontsize=12, fontweight='bold')
ax3.set_xlabel('Eje Z (Propagación)')
ax3.set_ylabel('E_x')
ax3.set_zlabel('E_y')

plt.tight_layout()
plt.show()
```

## 3. Ondas Periódicas, Armónicas y Estacionarias

### 3.1 Ondas Estacionarias y Armónicas
Cuando dos ondas de idéntica amplitud, frecuencia y longitud de onda viajan a través de un medio en sentidos opuestos, sus funciones de onda se superponen interfiriendo de forma constructiva y destructiva continuamente:
$$y_1(x, t) = A \cos(kx - \omega t)$$
$$y_2(x, t) = A \cos(kx + \omega t)$$

La superposición resultante $y(x, t) = y_1 + y_2$ aplicando identidades de suma es:
$$y(x, t) = 2A \sin(kx) \sin(\omega t)$$

Este patrón resultante no viaja espacialmente, sino que permanece fijo formando una **onda estacionaria**:
* **Nodos:** Puntos donde la oscilación es permanentemente nula ($kx = m\pi \implies x = m \frac{\lambda}{2}$).
* **Antinodos (Vientres):** Puntos de amplitud de oscilación máxima ($kx = (2m+1)\frac{\pi}{2} \implies x = (2m+1)\frac{\lambda}{4}$).

En una cuerda de longitud $L$ fija en ambos extremos, la condición de contorno de nodo en $x=0$ y $x=L$ restringe las longitudes de onda posibles a valores cuantizados:
$$L = n \frac{\lambda_n}{2} \implies \lambda_n = \frac{2L}{n}, \quad n \in \{1, 2, 3, \dots\}$$

Las frecuencias asociadas a cada modo o **armónicas** son múltiplos enteros de la **frecuencia fundamental** ($f_1$):
$$f_n = \frac{v}{\lambda_n} = n \left(\frac{v}{2L}\right) \implies f_n = n f_1$$

### 3.2 Ondas Electromagnéticas e Hipótesis de Planck
La luz se comporta como una onda electromagnética transversal compuesta por fotones. La energía de cada fotón ($E$) está relacionada de forma directa con su frecuencia espacial y temporal a través de la constante de Planck ($h \approx 6.626 \times 10^{-34}$ J·s):
$$E = h f = \frac{h c}{\lambda}$$

El **espectro electromagnético** se cataloga en función de este parámetro de longitud de onda:

| Región espectral | Rango de Longitud de Onda | Frecuencia típica (Hz) | Características físicas |
| :--- | :---: | :---: | :--- |
| **Radio** | $> 10^{-2}$ m | $10^4 - 10^8$ | No ionizante. Utilizada en telecomunicaciones. |
| **Microondas** | $10^{-2} - 10^{-5}$ m | $10^8 - 10^{12}$ | Calentamiento molecular rotacional. Radares y wifi. |
| **Infrarrojo** | $10^{-5} - 7 \times 10^{-7}$ m | $10^{12} - 10^{15}$ | Radiación térmica. Espectroscopía vibracional. |
| **Visible** | $700 - 400$ nm | $\sim 10^{15}$ | Región detectable por la retina humana. |
| **Ultravioleta** | $4 \times 10^{-7} - 10^{-8}$ m | $10^{15} - 10^{16}$ | Ionización molecular superficial. Esterilización. |
| **Rayos X** | $10^{-8} - 10^{-12}$ m | $10^{16} - 10^{20}$ | Alto poder de penetración atómica. Cristalografía. |
| **Rayos Gamma** | $< 10^{-12}$ m | $> 10^{20}$ | Radiación de origen nuclear y astrofísica violenta. |


```python
# Parámetros físicos de la cuerda
L_cuerda = 1.0  # Longitud (m)
x_cuerda = np.linspace(0, L_cuerda, 500)

fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)

# Graficar las 4 primeras armónicas (n = 1, 2, 3, 4)
for n in range(1, 5):
    # Perfil espacial de la armónica
    y_harm = np.sin(n * np.pi * x_cuerda / L_cuerda)
    
    ax = axes[n-1]
    ax.plot(x_cuerda, y_harm, 'b-', linewidth=2, label=f'Armónica n={n}')
    ax.plot(x_cuerda, -y_harm, 'b--', alpha=0.5) # Envolvente
    
    # Dibujar los nodos en el gráfico
    nodos_x = np.linspace(0, L_cuerda, n+1)
    ax.scatter(nodos_x, np.zeros_like(nodos_x), color='red', s=60, zorder=3, label='Nodos')
    
    ax.set_ylabel('Amplitud', fontsize=10)
    ax.set_ylim(-1.2, 1.2)
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(loc='upper right', fontsize=8)

axes[-1].set_xlabel('Posición x (m)', fontsize=11)
plt.suptitle('Primeros Cuatro Modos de Ondas Estacionarias (Armónicas)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. Energía, Potencia y Atenuación Geométrica

La energía mecánica transportada por un frente de onda es directamente proporcional al cuadrado de la amplitud de la perturbación ($A^2$) y al cuadrado de su frecuencia angular ($\omega^2$):
$$E \propto \omega^2 A^2 \propto f^2 A^2$$

La potencia ($P$) es la energía transmitida por unidad de tiempo ($P = E/t$, medida en Watts). La **Intensidad ($I$)** representa la potencia distribuida por unidad de área o superficie transversal al frente de onda:
$$I = \frac{P}{S} = \frac{E}{S \cdot t}$$

### 4.1 Atenuación Geométrica por Conservación de la Energía
Si asumimos un medio homogéneo y sin pérdidas disipativas, la energía total distribuida en un frente de onda en expansión se conserva. Sin embargo, a medida que la onda viaja, el área del frente de onda incrementa, distribuyendo la energía en una superficie mayor:

1. **Ondas Unidimensionales (1D):**
   El frente de onda se propaga sin ensancharse (ej. cuerda). El área de propagación es constante.
   $$S = \text{cte} \implies I = \text{cte} \implies A(r) \propto A_0$$
   La amplitud teórica se mantiene constante.

2. **Ondas Bidimensionales (2D):**
   El frente de onda se expande de forma circular en una superficie (ej. ondas en la superficie del agua).
   La circunferencia del frente a una distancia $r$ es $S = 2\pi r$.
   $$E_{frente} = \text{cte} \implies \frac{1}{2} \mu (2\pi r) A^2 \omega^2 = \text{cte} \implies r A^2 = \text{cte} \implies A(r) \propto \frac{1}{\sqrt{r}}$$
   La amplitud decrece de forma inversamente proporcional a la raíz cuadrada de la distancia $r$.

3. **Ondas Tridimensionales (3D):**
   El frente de onda se expande de forma esférica (ej. el sonido o la radiación electromagnética de una bombilla).
   El área de la esfera es $S = 4\pi r^2$.
   $$E_{frente} = \text{cte} \implies \frac{1}{2} \rho (4\pi r^2) A^2 \omega^2 = \text{cte} \implies r^2 A^2 = \text{cte} \implies A(r) \propto \frac{1}{r}$$
   La amplitud decrece linealmente con la distancia $r$. La intensidad disminuye cuadráticamente ($I \propto 1/r^2$), fenómeno conocido como la ley de la inversa del cuadrado.


```python
# Parámetros base
r_start = 0.5
r_end = 10.0
r = np.linspace(r_start, r_end, 500)
A0 = 1.0  # Amplitud inicial en r = 0.5

# 1D: Amplitud constante
A_1d = A0 * np.ones_like(r)

# 2D: A proportional to 1/sqrt(r)
A_2d = A0 * np.sqrt(r_start) / np.sqrt(r)

# 3D: A proportional to 1/r
A_3d = A0 * r_start / r

# Graficación de las curvas de decaimiento
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(r, A_1d, 'b-', linewidth=2.2, label='Propagación 1D ($A \\\\propto cte$)')
ax.plot(r, A_2d, 'g-', linewidth=2.2, label='Propagación 2D ($A \\\\propto 1/\\\\sqrt{r}$)')
ax.plot(r, A_3d, 'r-', linewidth=2.2, label='Propagación 3D ($A \\\\propto 1/r$)')

ax.set_title('Atenuación Geométrica de la Amplitud según la Dimensionalidad', fontsize=13, fontweight='bold')
ax.set_xlabel('Distancia al Foco Emisor r (m)', fontsize=11)
ax.set_ylabel('Amplitud Normalizada A', fontsize=11)
ax.set_xlim(r_start, r_end)
ax.set_ylim(0.0, 1.1)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(frameon=True, fontsize=10)

plt.tight_layout()
plt.show()
```

## 5. Efecto Doppler Clásico y Relativista

El **Efecto Doppler** es el cambio de frecuencia aparente de una onda debido al movimiento relativo entre la fuente emisora y el receptor u observador.

### 5.1 Efecto Doppler Clásico (Velocidades $v \ll c$)
En un medio de propagación acústica clásico, las velocidades se miden con respecto al medio inmóvil. La frecuencia recibida $f_o$ está gobernada por:
$$f_o = f_f \left( \frac{v \pm v_o}{v \mp v_f} \right)$$
Donde:
* $v$: Velocidad de propagación de la onda en el medio.
* $v_o$: Velocidad del observador.
* $v_f$: Velocidad de la fuente emisora.
* $f_f$: Frecuencia emitida en reposo.

**Criterio de Signos:**
* **Numerador ($v \pm v_o$):** Se usa ($+$) si el observador se mueve **hacia** la fuente (aumentando la frecuencia) y ($-$) si se aleja.
* **Denominador ($v \mp v_f$):** Se usa ($-$) si la fuente se mueve **hacia** el observador (acercándose y comprimiendo frentes de onda) y ($+$) si se aleja.

### 5.2 Efecto Doppler Relativista (Longitudinal en el Vacío)
Para la luz en el vacío, no hay un medio absoluto de referencia. El único parámetro relevante es la velocidad relativa $v$ entre fuente y receptor. Incorporando los efectos relativistas de dilatación del tiempo de la teoría especial de la relatividad (factor de Lorentz $\gamma = \frac{1}{\sqrt{1-\beta^2}}$), la frecuencia recibida es:
$$f_o = f_f \sqrt{\frac{1-\beta}{1+\beta}}$$
Donde $\beta = v/c$. Esta expresión asume que la fuente y el observador se están alejando. Si se estuvieran acercando, la fórmula invierte los signos:
$$f_o = f_f \sqrt{\frac{1+\beta}{1-\beta}}$$

### 5.3 Aplicaciones en Astronomía y Cosmología
Edwin Hubble utilizó el desplazamiento Doppler de las líneas espectrales del hidrógeno para calcular las velocidades de recesión de las galaxias lejanas. El parámetro adimensional de corrimiento al rojo $z$ se define como:
$$z = \frac{\lambda_o - \lambda_e}{\lambda_e} = \frac{f_e - f_o}{f_o}$$

Por lo tanto:
$$1 + z = \frac{\lambda_o}{\lambda_e} = \frac{f_e}{f_o}$$

Utilizando la expresión relativista:
$$1 + z = \sqrt{\frac{1+\beta}{1-\beta}}$$

* Si $z > 0$: Corrimiento al rojo (**redshift**), indicando que el objeto se está alejando.
* Si $z < 0$: Corrimiento al azul (**blueshift**), indicando que el objeto se está acercando.

Este fenómeno permitió formular la ley de expansión del Universo.


```python
# Visualización de los frentes de onda Doppler Clásico
# Fuente en movimiento de izquierda a derecha en un plano 2D
v_sound = 343.0  # m/s
v_source = 150.0 # m/s (Fuente subsónica)

# Posiciones de la fuente en 6 instantes de emisión pasados
t_instants = np.arange(0, 6)
x_emissions = v_source * t_instants
y_emissions = np.zeros_like(x_emissions)

t_current = 6.0
fig, ax = plt.subplots(figsize=(8, 8))

# Dibujar los frentes de onda circulares emitidos
for i, (xe, ye) in enumerate(zip(x_emissions, y_emissions)):
    dt_prop = t_current - t_instants[i]
    r_wave = v_sound * dt_prop
    circle = plt.Circle((xe, ye), r_wave, color='blue', fill=False, linestyle='-', linewidth=1.5, alpha=0.7)
    ax.add_patch(circle)
    # Marcar el centro de emisión
    ax.plot(xe, ye, 'rx', markersize=6)

# Posición actual de la fuente
x_current = v_source * t_current
ax.plot(x_current, 0, 'ro', markersize=10, label='Posición Actual de la Fuente')

# Configuración formal del plano
ax.set_title('Efecto Doppler: Compresión de Frentes de Onda por Fuente Móvil', fontsize=13, fontweight='bold')
ax.set_xlabel('Posición X (m)', fontsize=11)
ax.set_ylabel('Posición Y (m)', fontsize=11)
ax.set_xlim(-1000, 3000)
ax.set_ylim(-2000, 2000)
ax.set_aspect('equal')
ax.grid(True, linestyle=':', alpha=0.6)
ax.legend(frameon=True, loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Simulación de Corrimiento al Rojo Relativista en el Espectro de Hidrógeno

# Longitudes de onda de las líneas de Balmer en reposo (nm)
lambda_rest = {
    'H-alpha': 656.3,  # Rojo
    'H-beta': 486.1,   # Cian
    'H-gamma': 434.0,  # Azul
    'H-delta': 410.2   # Violeta
}

# Velocidad de recesión de la estrella (beta = v/c)
# Cambia beta para experimentar con corrimientos al rojo (>0) o al azul (<0)
beta = 0.15  # 15% de la velocidad de la luz

# Calcular las longitudes de onda observadas bajo el efecto relativista
lambda_obs = {}
for name, w in lambda_rest.items():
    # Fórmula relativista para alejamiento: lambda_o = lambda_e * sqrt((1+beta)/(1-beta))
    lambda_obs[name] = w * np.sqrt((1 + beta) / (1 - beta))

# Graficación del espectro óptico
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

# Espectro del Sol (En reposo)
ax1.set_xlim(380, 750)
ax1.set_ylim(0, 1)
# Renderizar un arcoíris de fondo continuo en el visible
gradient = np.linspace(0, 1, 1000)
ax1.imshow(np.vstack((gradient, gradient)), aspect='auto', cmap='rainbow', extent=[380, 750, 0, 1])

# Dibujar líneas de absorción (Líneas negras)
for name, w in lambda_rest.items():
    ax1.axvline(w, color='black', linewidth=3)
    ax1.text(w + 2, 0.7, name, rotation=90, color='black', fontsize=9, fontweight='bold')
ax1.set_title('Espectro de Hidrógeno en Reposo (Referencia)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Intensidad')
ax1.get_xaxis().set_visible(False)

# Espectro Observado (Desplazado)
ax2.set_xlim(380, 750)
ax2.set_ylim(0, 1)
ax2.imshow(np.vstack((gradient, gradient)), aspect='auto', cmap='rainbow', extent=[380, 750, 0, 1])

# Dibujar las líneas desplazadas
for name, w in lambda_obs.items():
    if 380 <= w <= 750:
        ax2.axvline(w, color='black', linewidth=3)
        ax2.text(w + 2, 0.7, f'{name} obs', rotation=90, color='black', fontsize=9, fontweight='bold')
z = np.sqrt((1 + beta) / (1 - beta)) - 1
ax2.set_title(f'Espectro Observado Desplazado ($z = {z:.4f}$ para $\\\\beta = {beta:.2f}$)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Longitud de onda (nm)', fontsize=11)
ax2.set_ylabel('Intensidad')

plt.tight_layout()
plt.show()
```

## 1. Repaso Matemático para Ondas: Trigonometría y Complejos

La descripción física de las ondas requiere una base matemática sólida en funciones armónicas y álgebra de complejos, que nos permita modelar la suma de múltiples señales ópticas o acústicas.

### 1.1 Identidades Trigonométricas de Suma y Producto
Cuando dos ondas coherentes se superponen (interferencia, Lección 3, Lección 9), sumamos sus funciones trigonométricas correspondientes. Para ello, son fundamentales las identidades de **suma a producto**:
$$\sin(X) + \sin(Y) = 2 \sin\left(\frac{X+Y}{2}\right) \cos\left(\frac{X-Y}{2}ight)$$
$$\cos(X) + \cos(Y) = 2 \cos\left(\frac{X+Y}{2}\right) \cos\left(\frac{X-Y}{2}ight)$$

Para analizar fenómenos como los **batidos** (beats, Lección 4) o la potencia transmitida, utilizamos las identidades de **reducción de potencia (ángulo medio)**:
$$\sin^2(\theta) = \frac{1 - \cos(2\theta)}{2}, \quad \cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}$$

### 1.2 La Fórmula de Euler y la Representación Compleja
En lugar de arrastrar funciones trigonométricas complejas (senos y cosenos), la física matemática simplifica las operaciones utilizando números complejos mediante la **Fórmula de Euler**:
$$e^{i\theta} = \cos(\theta) + i \sin(\theta)$$

Cualquier perturbación armónica real $\psi(x,t) = A \cos(kx - \omega t + \phi)$ puede representarse como la parte real de una función exponencial compleja:
$$\psi(x,t) = \text{Re}\left[ A e^{i(kx - \omega t + \phi)} \right] = \text{Re}\left[ \tilde{A} e^{i(kx - \omega t)} \right]$$
Donde $\tilde{A} = A e^{i\phi}$ es la **amplitud compleja**, la cual contiene tanto la información de la amplitud máxima ($A$) como el desfase de la onda ($\phi$). Esto simplifica notablemente las operaciones de derivación e integración, transformándolas en multiplicaciones algebraicas sencillas.

### 1.3 Ecuaciones Diferenciales del Movimiento Armónico Simple (MAS)
El Movimiento Armónico Simple (Lección 1) es el origen fundamental de las ondas. Su ecuación dinámica se rige por una ecuación diferencial lineal de segundo orden con coeficientes constantes:
$$\frac{d^2 x}{dt^2} + \omega_0^2 x = 0$$
Donde $\omega_0 = \sqrt{k/m}$ representa la frecuencia angular natural del sistema. La solución general de esta ecuación diferencial es de tipo armónico y se escribe como:
$$x(t) = A \cos(\omega_0 t + \phi)$$


## 2. La Ecuación de Onda y Ondas Armónicas Viajeras

Una onda es una perturbación que transporta energía y momento a través del espacio-tiempo sin necesidad de transferir materia.

### 2.1 La Ecuación Diferencial de Onda 1D
A partir de las leyes dinámicas (como la tensión en una cuerda o las ecuaciones de Maxwell, Lección 2, Lección 5), se deriva la **ecuación general de onda unidimensional**:
$$\frac{\partial^2 \psi}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 \psi}{\partial t^2}$$
Donde $\psi(x,t)$ representa la perturbación física (presión en el sonido, campo eléctrico en la luz) y $v$ es la velocidad de propagación de la onda.

### 2.2 Parámetros de una Onda Sinusoidal Viajera
Una de las soluciones más comunes e importantes de la ecuación de onda es la **onda armónica sinusoidal viajera**, la cual se desplaza hacia la derecha (eje $+x$) con la siguiente expresión física:
$$\psi(x, t) = A \cos(kx - \omega t + \phi)$$
Los parámetros físicos que la caracterizan y que el estudiante debe dominar son:
- **Amplitud ($A$):** La magnitud máxima de la perturbación respecto al equilibrio.
- **Longitud de onda ($\lambda$):** La distancia mínima entre dos puntos idénticos de la onda en un instante dado.
- **Número de onda ($k$):** Representa la frecuencia espacial de la onda y se define como $k = \frac{2\pi}{\lambda}$.
- **Periodo ($T$):** El tiempo necesario para que un ciclo completo pase por un punto fijo ($T = 1/f$).
- **Frecuencia angular ($\omega$):** La tasa de oscilación temporal expresada en radianes por segundo ($\omega = 2\pi f = \frac{2\pi}{T}$).
- **Velocidad de fase ($v$):** La velocidad a la que se propagan los frentes de onda, dada por la relación de dispersión lineal: $v = \lambda f = \frac{\omega}{k}$.
- **Fase inicial ($\phi$):** El desfase relativo en $x=0, t=0$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar estilo gráfico premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# Parámetros de la onda
A = 1.0          # Amplitud
lam = 4.0        # Longitud de onda (m)
k = 2 * np.pi / lam  # Número de onda
T = 2.0          # Periodo (s)
w = 2 * np.pi / T    # Frecuencia angular
v = lam / T      # Velocidad de fase (2.0 m/s)

x_vals = np.linspace(0, 12, 1000)

# =======================================================================
# 1. VISUALIZACIÓN DE UNA ONDA VIAJERA (Eje +x)
# =======================================================================
plt.figure(figsize=(10, 4))
t_snapshots = [0, T/4, T/2]
colores = ['royalblue', 'forestgreen', 'crimson']

for t, col in zip(t_snapshots, colores):
    psi = A * np.cos(k * x_vals - w * t)
    plt.plot(x_vals, psi, lw=2, color=col, label=rf'$t = {t:.2f}$ s')

plt.title(r'Propagación de una Onda Viajera Armónica (Hacia la Derecha): $\psi(x,t) = A\cos(kx - \omega t)$', fontsize=12, fontweight='bold')
plt.xlabel('Posición $x$ (m)')
plt.ylabel(r'Perturbación $\psi(x,t)$')
plt.ylim(-1.3, 1.3)
plt.legend(loc='upper right')
plt.show()

# =======================================================================
# 2. VISUALIZACIÓN DE UNA ONDA ESTACIONARIA (Superposición)
# =======================================================================
# Superposición de onda derecha e izquierda:
# psi_der = A * cos(k*x - w*t)
# psi_izq = A * cos(k*x + w*t)
# psi_sum = 2 * A * cos(k*x) * cos(w*t)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Instantes t = 0 (cos(w*t)=1) y t = T/4 (cos(w*t)=0)
instantes = [0, T/4]
labels_t = ['t = 0 (Amplitud Máxima)', 't = T/4 (Nodos en Cero)']

for ax, t_val, lbl in zip(axes, instantes, labels_t):
    psi_der = A * np.cos(k * x_vals - w * t_val)
    psi_izq = A * np.cos(k * x_vals + w * t_val)
    psi_est = 2 * A * np.cos(k * x_vals) * np.cos(w * t_val)
    
    ax.plot(x_vals, psi_der, 'b--', alpha=0.5, label='Onda Derecha (+x)')
    ax.plot(x_vals, psi_izq, 'g--', alpha=0.5, label='Onda Izquierda (-x)')
    ax.plot(x_vals, psi_est, 'r-', lw=2.5, label='Onda Estacionaria (Suma)')
    
    # Marcar los nodos teóricos: cos(k*x) = 0 => k*x = (2n+1)*pi/2 => x = (2n+1)*lam/4
    # Para lambda = 4: x = (2n+1)*1 = 1, 3, 5, 7, 9, 11...
    nodos_x = np.array([1, 3, 5, 7, 9, 11])
    ax.scatter(nodos_x, np.zeros_like(nodos_x), color='black', s=60, zorder=5, label='Nodos (N)')
    
    # Marcar antinodos teóricos: cos(k*x) = +-1 => k*x = n*pi => x = n*lam/2
    # Para lambda = 4: x = n*2 = 0, 2, 4, 6, 8, 10, 12...
    antinodos_x = np.array([0, 2, 4, 6, 8, 10, 12])
    ax.scatter(antinodos_x, 2 * A * np.cos(k * antinodos_x) * np.cos(w * t_val), color='orange', edgecolor='black', s=60, zorder=5, label='Antinodos (A)')
    
    ax.set_title(lbl, fontsize=12, fontweight='bold')
    ax.set_xlabel('Posición $x$ (m)')
    ax.set_ylabel(r'Perturbación $\psi(x,t)$')
    ax.set_ylim(-2.3, 2.3)
    ax.legend(loc='upper right')

plt.suptitle('Generación de una Onda Estacionaria y Estructura de Nodos/Antinodos', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 3. El Espectro Electromagnético y el Rango Visible

La luz es una onda electromagnética transversal compuesta por campos eléctricos y magnéticos oscilantes mutuamente perpendiculares (Lección 5). El conjunto de todas las ondas electromagnéticas se organiza según su longitud de onda o frecuencia en el **Espectro Electromagnético**.

### 3.1 Bandas del Espectro Electromagnético
El espectro se clasifica comúnmente en las siguientes regiones (ordenadas de menor a mayor energía):
1. **Ondas de Radio ($\lambda > 1\text{ m}$):** Utilizadas en telecomunicaciones.
2. **Microondas ($1\text{ mm} < \lambda < 1\text{ m}$):** Utilizadas en radares y hornos de microondas.
3. **Infrarrojo ($750\text{ nm} < \lambda < 1\text{ mm}$):** Radiación térmica de cuerpos a temperatura ambiente.
4. **Visible ($380\text{ nm} \le \lambda \le 750\text{ nm}$):** La única banda estrecha que puede detectar el ojo humano.
5. **Ultravioleta ($10\text{ nm} < \lambda < 380\text{ nm}$):** Radiación ionizante de baja energía emitida por el Sol.
6. **Rayos X ($0.01\text{ nm} < \lambda < 10\text{ nm}$):** Alta energía, capaces de penetrar tejidos corporales blandos.
7. **Rayos Gamma ($\lambda < 0.01\text{ nm}$):** Emitidos por desintegraciones nucleares y cataclismos astrofísicos.

### 3.2 El Espectro Visible y su Mapeo Cromático
Dentro del espectro visible, el ojo humano percibe diferentes longitudes de onda como colores específicos. En física computacional, a menudo queremos simular el color exacto de la luz láser en función de su longitud de onda. Para ello, utilizamos un algoritmo de conversión física que asocia a cada longitud de onda $\lambda \in [380, 750]\text{ nm}$ un vector de color RGB (Red, Green, Blue) de acuerdo con la respuesta fisiológica de los fotorreceptores (conos) del ojo humano.


```python
def lambda_to_rgb(wl_nm):
    """
    Algoritmo físico de Bruton para mapear una longitud de onda en nanómetros
    a sus coeficientes de color RGB (normalizados entre 0 y 1).
    """
    w = float(wl_nm)
    # Inicializar componentes
    R, G, B = 0.0, 0.0, 0.0
    
    # Definición de color por bandas del espectro visible
    if w >= 380 and w < 440:
        R = -(w - 440) / (440 - 380)
        G = 0.0
        B = 1.0
    elif w >= 440 and w < 490:
        R = 0.0
        G = (w - 440) / (490 - 440)
        B = 1.0
    elif w >= 490 and w < 510:
        R = 0.0
        G = 1.0
        B = -(w - 510) / (510 - 490)
    elif w >= 510 and w < 580:
        R = (w - 510) / (580 - 510)
        G = 1.0
        B = 0.0
    elif w >= 580 and w < 645:
        R = 1.0
        G = -(w - 645) / (645 - 580)
        B = 0.0
    elif w >= 645 and w <= 750:
        R = 1.0
        G = 0.0
        B = 0.0
        
    # Atenuación de intensidad cerca de los límites del ojo (380-420 nm y 700-750 nm)
    if w >= 380 and w < 420:
        factor = 0.3 + 0.7 * (w - 380) / (420 - 380)
    elif w >= 420 and w < 701:
        factor = 1.0
    elif w >= 701 and w <= 750:
        factor = 0.3 + 0.7 * (750 - w) / (750 - 701)
    else:
        factor = 0.0
        
    return R * factor, G * factor, B * factor

# Generar el mapeo cromático continuo para el espectro visible (380 a 750 nm)
wls = np.linspace(380, 750, 600)
rgb_colors = np.array([lambda_to_rgb(wl) for wl in wls])

# Graficar la barra del espectro visible
plt.figure(figsize=(12, 2.5))
# Reshape a (1, N_wavelengths, 3) para usar con imshow
spectrum_img = rgb_colors.reshape(1, -1, 3)
plt.imshow(spectrum_img, extent=[380, 750, 0, 1], aspect='auto')
plt.title('Espectro Electromagnético Visible y sus Colores Físicos Asociados', fontsize=12, fontweight='bold')
plt.xlabel(r'Longitud de Onda $\lambda$ (nm)')
plt.yticks([])  # Quitar eje Y para simular una barra de color
plt.grid(False) # Quitar rejilla
plt.show()
```

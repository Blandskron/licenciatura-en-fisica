## 1. Ondas de Sonido y sus Características

Las **ondas sonoras** son perturbaciones mecánicas longitudinales que se propagan en un medio elástico. En una onda longitudinal, el desplazamiento de las partículas del medio es paralelo a la dirección de propagación de la onda. A nivel microscópico, una fuente vibrante (como una cuerda de guitarra o las cuerdas vocales) empuja las partículas de aire adyacentes hacia adelante y hacia atrás. Esto genera alternadamente zonas de alta densidad y presión, llamadas **compresiones**, y zonas de baja densidad y presión, llamadas **rarefacciones**.

### 1.1 Características Físicas y Psicoacústicas
El sonido se puede describir físicamente mediante los siguientes parámetros:
* **Frecuencia ($f$):** Número de oscilaciones completas por segundo, medido en Hertz (Hz). Determina la propiedad psicoacústica del **tono** (si un sonido es grave o agudo). El oído humano es sensible al rango entre $20\text{ Hz}$ y $20\,000\text{ Hz}$. Las ondas por debajo de este rango se denominan **infrasonidos** y las que están por encima, **ultrasonidos**.
* **Amplitud ($A$ o $\Delta p_{\text{max}}$):** Magnitud de la variación de presión respecto al equilibrio. Determina el **volumen** o la intensidad del sonido. Frecuentemente se mide en la escala logarítmica de decibelios (dB).
* **Longitud de onda ($\lambda$):** Distancia física entre dos compresiones o rarefacciones consecutivas. Se relaciona con la velocidad de propagación $v$ y la frecuencia $f$ mediante:
  $$\lambda = \frac{v}{f}$$
* **Velocidad de propagación ($v$):** Velocidad con la que viaja la perturbación a través del medio. Depende de las propiedades elásticas e inerciales del material (ej. $\sim 343\text{ m/s}$ en el aire a $20^\circ\text{C}$).
* **Dirección de propagación:** El sonido se propaga en línea recta en un medio homogéneo e isótropo, expandiéndose de forma esférica tridimensional desde una fuente puntual.

### 1.2 Formulación Matemática
Podemos describir una onda sonora armónica unidimensional en términos del desplazamiento molecular $s(x, t)$ respecto a su posición de equilibrio $x$:
$$s(x, t) = s_{\text{max}} \cos(kx - \omega t)$$
Donde $s_{\text{max}}$ es el desplazamiento máximo, $k = 2\pi/\lambda$ es el número de onda y $\omega = 2\pi f$ es la frecuencia angular.

La variación de presión acústica $\Delta p(x, t)$ respecto a la presión de equilibrio del fluido está relacionada con el gradiente de desplazamiento a través del módulo de compresibilidad volumétrico $B$ del medio:
$$\Delta p(x, t) = -B \frac{\partial s}{\partial x} = B k s_{\text{max}} \sin(kx - \omega t) = \Delta p_{\text{max}} \sin(kx - \omega t)$$
Donde la amplitud de presión es:
$$\Delta p_{\text{max}} = B k s_{\text{max}} = \rho v^2 k s_{\text{max}} = \rho v \omega s_{\text{max}}$$
Esto demuestra que la onda de presión está desfasada $90^\circ$ ($\pi/2$ radianes) respecto a la onda de desplazamiento de las partículas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar el estilo visual para los gráficos científicos
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Parámetros de la onda sonora
wavelength = 2.0  # Longitud de onda (m)
k = 2 * np.pi / wavelength
amplitude_displacement = 0.15  # Amplitud del desplazamiento (m)
num_particles = 1500
length_tube = 6.0

# Posiciones de equilibrio de las partículas en un tubo 2D
x_equilibrium = np.random.uniform(0, length_tube, num_particles)
y_particles = np.random.uniform(-0.5, 0.5, num_particles)

# Desplazamiento longitudinal s(x, t) a t = 0
displacement = amplitude_displacement * np.cos(k * x_equilibrium)
x_current = x_equilibrium + displacement

# Crear el gráfico comparativo
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 7), sharex=True,
                               gridspec_kw={'height_ratios': [2, 1.5]})

# Panel 1: Distribución microscópica de las partículas (compresión y rarefacción)
ax1.scatter(x_current, y_particles, color='royalblue', s=8, alpha=0.7, label='Moléculas de Aire')
ax1.set_title('Representación Microscópica y Macroscópica de una Onda Sonora', fontsize=13, fontweight='bold')
ax1.set_ylabel('Eje Y (m)', fontsize=11)
ax1.set_ylim(-0.8, 0.8)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True, loc='upper right')

# Panel 2: Onda macroscópica de presión acústica Delta P(x)
x_axis = np.linspace(0, length_tube, 500)
pressure_wave = np.sin(k * x_axis)
ax2.plot(x_axis, pressure_wave, color='crimson', linewidth=2.5, label='Presión Acústica $\\Delta P(x)$')
ax2.axhline(0, color='black', linewidth=1, linestyle='--')
ax2.set_xlabel('Posición a lo largo del tubo X (m)', fontsize=11)
ax2.set_ylabel('$\\Delta P / \\Delta P_{max}$', fontsize=11)
ax2.set_ylim(-1.2, 1.2)
ax2.grid(True, linestyle=':', alpha=0.5)
ax2.legend(frameon=True, loc='upper right')

# Añadir anotaciones para relacionar densidad de partículas con presión
# Compresión corresponde a máximos locales y rarefacción a mínimos
ax1.annotate('Compresión (Alta densidad)', xy=(1.0, 0.6), xytext=(0.3, 0.9),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))
ax1.annotate('Rarefacción (Baja densidad)', xy=(2.0, 0.6), xytext=(2.2, 0.9),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

plt.tight_layout()
plt.show()
```

## 2. Reflexión y Refracción del Sonido

Al igual que otras ondas físicas, cuando las ondas sonoras encuentran una discontinuidad en el medio o cambian de medio de propagación, experimentan fenómenos de reflexión y refracción.

### 2.1 Reflexión de Ondas Sonoras
La **reflexión** ocurre cuando una onda choca contra una interfaz y cambia su dirección, regresando al medio original. Sigue la ley de reflexión geométrica clásica: el ángulo de incidencia $\theta_i$ con respecto a la normal de la superficie es igual al ángulo de reflexión $\theta_r$:
$$\theta_i = \theta_r$$
La fracción de energía sonora reflejada frente a la transmitida depende de la diferencia en la **impedancia acústica** ($Z$) de los dos medios, definida como la resistencia al paso de las ondas mecánicas:
$$Z = \rho \cdot v$$
Si la diferencia de impedancias es muy grande (como entre el aire y una pared de concreto), casi toda la energía sonora es reflejada. Esto causa fenómenos como el **eco** (cuando la reflexión se percibe separada en el tiempo, típicamente con retardos mayores a $0.1\text{ s}$) y la **reverberación** (superposición de reflexiones múltiples y continuas).

### 2.2 Refracción del Sonido
La **refracción** es el cambio de dirección que experimenta la onda al pasar de un punto a otro con diferente velocidad de propagación. Un caso sumamente interesante ocurre en la atmósfera debido a gradientes de temperatura verticales:
1. **Durante el día (Atmósfera Normal):** El suelo caliente calienta la capa inferior de aire, por lo que la velocidad del sonido es mayor cerca de la superficie y menor a mayor altitud. Dado que la parte inferior de los frentes de onda viaja más rápido, el frente de onda se curva **hacia arriba**, dificultando la audición de sonidos lejanos a ras de suelo.
2. **Durante la noche o mañana (Inversión Térmica):** El aire cerca del suelo está frío (por ejemplo, sobre un lago) mientras que las capas superiores están más templadas por radiación solar inicial. La velocidad del sonido es menor abajo y mayor arriba, lo que causa que los frentes de onda se curven **hacia abajo**. Esto confina la energía sonora cerca de la superficie, actuando como una guía de onda y permitiendo escuchar voces o ruidos a grandes distancias sobre el agua.


```python
# Simulación de trayectoria de rayos de sonido bajo gradientes térmicos
# Para un gradiente de velocidad v(y) = v_0 + g*y, los rayos describen trayectorias circulares

def calcular_rayo(angle_degrees, gradient, v0=340.0, num_steps=200, dt=0.005):
    # Convertir ángulo de salida con respecto a la horizontal
    theta = np.radians(angle_degrees)
    x = [0.0]
    y = [0.0]  # Origen
    
    # Componentes de la dirección del rayo
    vx = v0 * np.cos(theta)
    vy = v0 * np.sin(theta)
    
    curr_x, curr_y = 0.0, 0.0
    for _ in range(num_steps):
        # Velocidad local según la altitud y
        v_local = v0 + gradient * curr_y
        if v_local <= 100.0:  # Límite físico inferior
            break
            
        # Usar la ley de Snell en forma diferencial para actualizar el ángulo
        # Para v(y) = v0 + g*y: d(theta)/dt = -g * cos(theta)
        angle_curr = np.arctan2(vy, vx)
        dtheta = -gradient * np.cos(angle_curr) * dt
        angle_new = angle_curr + dtheta
        
        # Actualizar componentes de velocidad
        vx = v_local * np.cos(angle_new)
        vy = v_local * np.sin(angle_new)
        
        # Actualizar posición
        curr_x += vx * dt
        curr_y += vy * dt
        x.append(curr_x)
        y.append(curr_y)
        
    return np.array(x), np.array(y)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
angles_test = [10, 20, 30, 40, 50]  # Ángulos de emisión de rayos (grados)

# 1. Atmósfera Diurna (Gradiente negativo: calor abajo, frío arriba)
g_day = -15.0  # s^-1
for angle in angles_test:
    rx, ry = calcular_rayo(angle, g_day)
    ax1.plot(rx, ry, 'b-', linewidth=2)
ax1.set_title('Día Normal: Gradiente Térmico Negativo (Sonido sube)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Distancia Horizontal X (m)', fontsize=11)
ax1.set_ylabel('Altitud Y (m)', fontsize=11)
ax1.set_xlim(0, 300)
ax1.set_ylim(0, 100)
ax1.grid(True, linestyle=':', alpha=0.6)

# 2. Inversión Térmica Nocturna (Gradiente positivo: frío abajo, templado arriba)
g_night = 15.0  # s^-1
for angle in angles_test:
    rx, ry = calcular_rayo(angle, g_night)
    ax2.plot(rx, ry, 'r-', linewidth=2)
ax2.set_title('Inversión Térmica: Gradiente Positivo (Sonido baja y se propaga)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Distancia Horizontal X (m)', fontsize=11)
ax2.set_xlim(0, 300)
ax2.set_ylim(0, 100)
ax2.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

## 3. Derivación de la Velocidad del Sonido en Gases Ideales

Para deducir de manera formal y cuantitativa la velocidad de propagación del sonido en un fluido gaseoso, consideramos el paso de una perturbación acústica unidimensional a través de un tubo de sección transversal $A$. Analicemos las propiedades físicas de un elemento de fluido utilizando principios fundamentales de conservación.

### 3.1 Conservación de la Masa (Continuidad)
Consideremos un volumen de control en el tubo. Al paso de la onda de sonido, las partículas entran por un extremo con densidad $\rho$ y velocidad $v$. Al salir por el otro extremo, sus propiedades han variado infinitesimalmente a $\rho + d\rho$ y $v + dv$. El flujo de masa entrante por unidad de tiempo es igual al flujo saliente para mantener la continuidad:
$$\rho A v = (\rho + d\rho) A (v + dv)$$
Dividiendo por el área $A$ y desarrollando el producto del lado derecho:
$$\rho v = \rho v + \rho \, dv + v \, d\rho + d\rho \, dv$$
Como el producto de dos diferenciales infinitesimales es de segundo orden ($d\rho \, dv \approx 0$), obtenemos la **ecuación de continuidad diferencial**:
$$0 = \rho \, dv + v \, d\rho \implies v \, d\rho = -\rho \, dv \implies \frac{dv}{d\rho} = -\frac{v}{\rho} \quad \text{(Ecuación 1)}$$

### 3.2 Conservación del Momento Lineal (Segunda Ley de Newton)
Analicemos ahora las fuerzas dinámicas sobre el elemento de fluido. La presión en la cara izquierda del elemento es $P$, lo que produce una fuerza hacia la derecha $P A$. En la cara derecha, la presión es $P + dP$, generando una fuerza hacia la izquierda de $(P + dP)A$. La fuerza neta es la diferencia de presiones a lo largo del volumen de control:
$$F_n = P A - (P + dP)A = -A \, dP$$

Aplicando la segunda ley de Newton ($F_n = m \cdot a$):
$$-A \, dP = (dm) \frac{dv}{dt}$$
Dado que la masa del elemento es $dm = \rho A dx$ y sabiendo que la velocidad de propagación de la perturbación se relaciona con el diferencial espacial mediante $dx/dt = v$, reescribimos la ecuación como:
$$-A \, dP = (\rho A dx) \frac{dv}{dt} = \rho A \left(\frac{dx}{dt}\right) dv = \rho A v \, dv$$
Cancelando el área $A$ en ambos lados:
$$-dP = \rho v \, dv \implies v \, dv = -\frac{dP}{\rho} \quad \text{(Ecuación 2)}$$

### 3.3 Combinación de Ecuaciones
Sustituyendo el diferencial de velocidad de la Ecuación 1 ($dv = -\frac{v}{\rho} d\rho$) en la Ecuación 2:
$$v \left( -\frac{v}{\rho} d\rho \right) = -\frac{dP}{\rho} \implies -v^2 \frac{d\rho}{\rho} = -\frac{dP}{\rho} \implies v^2 = \frac{dP}{d\rho}$$
Extrayendo la raíz cuadrada de ambos lados, obtenemos la relación general para la velocidad del sonido en cualquier medio fluido:
$$v = \sqrt{\frac{dP}{d\rho}} \quad \text{(Ecuación de Newton-Laplace)}$$

### 3.4 El Límite Termodinámico Adiabático
Isaac Newton dedujo inicialmente esta relación asumiendo un proceso isotérmico ($T = \text{constante}$). Sin embargo, esto producía un error del $15\%$ con respecto a los experimentos. Pierre-Simon Laplace corrigió este modelo explicando que las compresiones y rarefacciones son tan rápidas que el calor no tiene tiempo de difundirse. Por lo tanto, el proceso de propagación del sonido es **adiabático**.

Para un gas ideal en un proceso adiabático reversible, se cumple:
$$P V^\gamma = \text{constante}$$
Donde $\gamma$ es el coeficiente adiabático del gas. Dado que la densidad molecular es inversamente proporcional al volumen ($\rho \propto 1/V$), expresamos la presión en términos de la densidad:
$$P \left(\frac{1}{\rho}\right)^\gamma = C \implies P = C \rho^\gamma$$
Donde $C$ es una constante. Derivando respecto a la densidad $\rho$:
$$\frac{dP}{d\rho} = C \gamma \rho^{\gamma - 1} = \gamma \left( C \rho^\gamma \right) \frac{1}{\rho} = \gamma \frac{P}{\rho}$$
Sustituyendo esta derivada en la ecuación de Newton-Laplace, obtenemos la velocidad del sonido adiabática:
$$v = \sqrt{\frac{\gamma P}{\rho}} \quad \text{(Ecuación 3)}$$

### 3.5 Expresión en Función de la Temperatura
Utilizando la ley de los gases ideales ($P V = nRT$):
$$P = \frac{nRT}{V} = \frac{m}{M} \frac{RT}{V} = \left(\frac{m}{V}\right) \frac{RT}{M} = \frac{\rho RT}{M}$$
Donde $M$ es la masa molar del gas y $R = 8.314\text{ J/(mol K)}$ es la constante universal de los gases. Sustituyendo $P/\rho = RT/M$ en la Ecuación 3:
$$v = \sqrt{\frac{\gamma RT}{M}}$$

Para el aire atmosférico normal (gas diatómico mayoritariamente):
* $\gamma \approx 1.40$
* Masa molar promedio del aire: $M \approx 0.029\text{ kg/mol}$
* A la temperatura de congelación del agua ($T_0 = 273.15\text{ K} = 0^\circ\text{C}$):
  $$v_0 = \sqrt{\frac{1.40 \times 8.314 \times 273.15}{0.029}} \approx 331.3\text{ m/s}$$
* Podemos expresar la velocidad a cualquier temperatura absoluta $T$ en Kelvin en relación con esta constante:
  $$v(T) = v_0 \sqrt{\frac{T}{273.15}} = 331.3 \sqrt{\frac{T}{273.15}}\text{ m/s}$$


```python
# Comparación gráfica de la velocidad del sonido teórica (adiabática) y su aproximación lineal

# Constantes del aire
gamma = 1.40
R = 8.314  # J/(mol K)
M_air = 0.029  # kg/mol

# Rango de temperaturas en Celsius y Kelvin
T_celsius = np.linspace(-20, 40, 100)
T_kelvin = T_celsius + 273.15

# Velocidad del sonido exacta (Teoría Adiabática)
v_exact = np.sqrt((gamma * R * T_kelvin) / M_air)

# Velocidad del sonido aproximada (Aproximación Lineal clásica para rango ambiente)
v_approx = 331.3 + 0.606 * T_celsius

# Calcular el error relativo porcentual
error_rel = np.abs(v_exact - v_approx) / v_exact * 100

# Graficación con subplots para la curva de velocidad y el perfil de error
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True,
                               gridspec_kw={'height_ratios': [3, 1]})

# Gráfico 1: Curvas de velocidad
ax1.plot(T_celsius, v_exact, 'b-', linewidth=2.5, label='Velocidad Adiabática Exacta $v = \\sqrt{\\gamma R T / M}$') 
ax1.plot(T_celsius, v_approx, 'r--', linewidth=2, label='Aproximación Lineal $v \\approx 331.3 + 0.606 T$')
ax1.set_title('Velocidad del Sonido en el Aire vs Temperatura', fontsize=13, fontweight='bold')
ax1.set_ylabel('Velocidad del Sonido $v$ (m/s)', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True, loc='upper left')

# Añadir anotación con el valor a 20 grados Celsius
v_20 = np.sqrt((gamma * R * (20 + 273.15)) / M_air)
ax1.scatter(20, v_20, color='black', zorder=5)
ax1.annotate(f'{v_20:.1f} m/s a 20°C', xy=(20, v_20), xytext=(23, v_20 - 5),
             arrowprops=dict(facecolor='black', shrink=0.08, width=0.5, headwidth=5))

# Gráfico 2: Error relativo porcentual
ax2.plot(T_celsius, error_rel, 'g-', linewidth=2, label='Error Relativo %')
ax2.set_xlabel('Temperatura ($^\\circ$C)', fontsize=11)
ax2.set_ylabel('Error (%)', fontsize=11)
ax2.grid(True, linestyle=':', alpha=0.5)
ax2.legend(frameon=True, loc='upper right')
ax2.set_ylim(-0.05, 0.4)

plt.tight_layout()
plt.show()
```

## 4. Interferencia y Batidos (Beats)

La superposición lineal de ondas acústicas da lugar a interesantes patrones espaciales y temporales.

### 4.1 Interferencia Coherente (Patrón Espacial)
Consideremos dos fuentes puntuales coherentes (misma frecuencia $f$, longitud de onda $\lambda$ y fase inicial) que interfieren en un punto $P$. Las distancias de las fuentes a $P$ son $x_1$ y $x_2$.
Las señales en $P$ vienen dadas por:
$$F_1 = A \sin(\omega t - k x_1), \quad F_2 = A \sin(\omega t - k x_2)$$
Sumando las perturbaciones y aplicando la identidad de suma de senos:
$$F_T = F_1 + F_2 = \left[ 2A \cos\left( \frac{k (x_2 - x_1)}{2} \right) \right] \sin\left( \omega t - k \frac{x_1 + x_2}{2} \right)$$
La amplitud de la onda resultante es:
$$A_T = 2A \cos\left( \frac{k \Delta x}{2} \right) = 2A \cos\left( \frac{\pi \Delta x}{\lambda} \right) = 2A \cos\left( \frac{\Delta \phi}{2} \right)$$
Donde $\Delta x = x_2 - x_1$ es la diferencia de camino y $\Delta \phi = k \Delta x$ es la diferencia de fase.

* **Interferencia Constructiva (Máxima Intensidad):**
  $$\frac{\pi \Delta x}{\lambda} = n\pi \implies \Delta x = n\lambda, \quad n \in \{0, 1, 2, \dots\} \implies A_T = \pm 2A$$
* **Interferencia Destructiva (Silencio o Amplitud Nula):**
  $$\frac{\pi \Delta x}{\lambda} = (2n + 1)\frac{\pi}{2} \implies \Delta x = (2n + 1)\frac{\lambda}{2}, \quad n \in \{0, 1, 2, \dots\} \implies A_T = 0$$

### 4.2 Interferencia No Coherente: Batidos (Pulsaciones)
Cuando se superponen dos ondas de la misma amplitud pero con **frecuencias ligeramente diferentes** $f_1$ y $f_2$, las ondas no son coherentes debido a que su diferencia de fase cambia constantemente en el tiempo. Analizando el fenómeno en la posición $x = 0$:
$$F_1 = A \sin(2\pi f_1 t), \quad F_2 = A \sin(2\pi f_2 t)$$
La superposición resulta en:
$$F_T(t) = \left[ 2A \cos\left(2\pi \frac{f_1 - f_2}{2} t\right) \right] \sin\left(2\pi \frac{f_1 + f_2}{2} t\right)$$
Podemos interpretar este resultado como una oscilación rápida a la frecuencia promedio $f_{\text{prom}} = \frac{f_1 + f_2}{2}$ cuya amplitud está modulada lentamente en el tiempo por el término envolvente:
$$A_{\text{env}}(t) = 2A \cos\left(2\pi \frac{f_1 - f_2}{2} t\right)$$
El oído humano percibe la intensidad sonora (proporcional al cuadrado de la amplitud). El valor absoluto de la envolvente tiene dos máximos por cada ciclo de la función coseno. Por lo tanto, la frecuencia con la que se escuchan los pulsos o batidos (frecuencia de batido, $f_{\text{beat}}$) es exactamente el doble de la frecuencia del término modulador, es decir:
$$f_{\text{beat}} = 2 \left( \frac{|f_1 - f_2|}{2} \right) = |f_1 - f_2|$$
Este fenómeno es el principio físico utilizado para la afinación de instrumentos musicales de cuerda y viento, donde el afinador ajusta la tensión hasta que la frecuencia de batido percibida disminuye a cero.


```python
# Simulación temporal del fenómeno de batidos (pulsaciones)

f1 = 200.0  # Frecuencia de la primera onda (Hz)
f2 = 220.0  # Frecuencia de la segunda onda (Hz) - ligeramente diferente
A = 1.0     # Amplitud de cada onda

f_prom = (f1 + f2) / 2.0
f_beat = np.abs(f1 - f2)

# Generar vector de tiempo para visualizar unos 4 ciclos de batido (T_beat = 1/20 Hz = 0.05 s)
t = np.linspace(0, 0.20, 2000)

wave1 = A * np.sin(2 * np.pi * f1 * t)
wave2 = A * np.sin(2 * np.pi * f2 * t)
wave_resultant = wave1 + wave2

# Envolvente de modulación lenta
envelope = 2 * A * np.cos(2 * np.pi * (f1 - f2) / 2.0 * t)

plt.figure(figsize=(12, 5))
plt.plot(t * 1000, wave_resultant, color='royalblue', label='Señal Superpuesta (Batidos)')
plt.plot(t * 1000, envelope, 'r--', label='Envolvente Superior $\\pm 2A \\cos(\\Delta \\omega t / 2)$', alpha=0.85, linewidth=1.5)
plt.plot(t * 1000, -envelope, 'r--', alpha=0.85, linewidth=1.5)

plt.title(f'Fenómeno de Batidos Acústicos ($f_1 = {f1}$ Hz, $f_2 = {f2}$ Hz) $\\rightarrow f_{{beat}} = {f_beat}$ Hz', fontsize=13, fontweight='bold')
plt.xlabel('Tiempo (ms)', fontsize=11)
plt.ylabel('Amplitud Resultante', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper right')
plt.xlim(0, t[-1]*1000)
plt.show()
```

## 5. Ondas Sonoras Estacionarias y la Base de la Música

Las ondas estacionarias se producen por la superposición de dos ondas viajeras de la misma frecuencia y amplitud que se propagan en direcciones opuestas dentro de un espacio confinado, como una columna de aire en un tubo.

### 5.1 Tubo Abierto en Ambos Extremos
En un tubo con ambos extremos abiertos, el aire en las aberturas está en contacto con la atmósfera exterior y puede oscilar libremente (máximo desplazamiento molecular). Esto impone **nodos de presión** en ambos extremos ($x = 0$ y $x = L$):
$$\Delta p(0, t) = 0 \quad \text{y} \quad \Delta p(L, t) = 0$$
Esto obliga a que la longitud del tubo sea un múltiplo entero de medias longitudes de onda:
$$L = n \frac{\lambda_n}{2} \implies \lambda_n = \frac{2L}{n}, \quad n \in \{1, 2, 3, \dots\}$$
Las frecuencias de resonancia permitidas (armónicos) son:
$$f_n = n \left( \frac{v}{2L} \right), \quad n \in \{1, 2, 3, \dots\}$$
Donde $n=1$ corresponde al armónico fundamental (primer armónico). Un tubo abierto genera la serie completa de armónicos enteros.

### 5.2 Tubo Cerrado en Un Extremo (Cerrado-Abierto)
En el extremo cerrado, las moléculas de aire no pueden desplazarse debido a la pared rígida (nodo de desplazamiento), lo que produce una acumulación de presión (**antinodo de presión**). En el extremo abierto se mantiene el nodo de presión:
* Extremo cerrado ($x = 0$): Antinodo de presión (gradiente nulo de presión: $\left. \frac{\partial \Delta p}{\partial x} \right|_{x=0} = 0$).
* Extremo abierto ($x = L$): Nodo de presión ($\Delta p(L, t) = 0$).
Esto obliga a que la longitud del tubo sea un múltiplo impar de cuartos de longitud de onda:
$$L = (2n - 1) \frac{\lambda_n}{4} \implies \lambda_n = \frac{4L}{2n - 1}, \quad n \in \{1, 2, 3, \dots\}$$
Las frecuencias de resonancia permitidas son:
$$f_n = (2n - 1) \left( \frac{v}{4L} \right), \quad n \in \{1, 2, 3, \dots\}$$
Un tubo cerrado solo genera armónicos **impares** (1.º, 3.º, 5.º, etc.), lo que altera radicalmente la riqueza de su espectro acústico.

### 5.3 La Física de la Música y la Escala Temperada
En la música occidental, el espectro acústico se organiza en **octavas**. Una octava representa una duplicación de la frecuencia fundamental (ej. La 440 Hz a La 880 Hz). Entre dos notas de octavas sucesivas se definen 12 intervalos musicales llamados **semitonos**.
Para que la distancia tonal percibida entre cualquier par de semitonos adyacentes sea constante, la relación de frecuencias debe ser geométrica. Por tanto, la frecuencia de una nota se obtiene multiplicando la anterior por una constante de proporcionalidad $k$:
$$f_{m+1} = k \cdot f_m \implies f_{m+12} = k^{12} \cdot f_m = 2 f_m \implies k = \sqrt[12]{2} \approx 1.059463$$
Esta escala se conoce como la **escala temperada**.

### 5.4 Timbre y Síntesis Armónica
Dos instrumentos diferentes tocando la misma nota fundamental (ej. un Do a 261.6 Hz) suenan de forma distinta debido al **timbre**. El timbre está determinado por la presencia y la amplitud relativa de los armónicos superiores que acompañan a la frecuencia fundamental. Matemáticamente, cualquier onda periódica compleja se puede reconstruir usando series de Fourier como una suma infinita de senos y cosenos armónicos:
$$F(t) = \sum_{n=1}^{\infty} A_n \sin(n \omega_0 t + \phi_n)$$


```python
# 1. Simulación de los perfiles de presión acústica en tubos
L = 1.0  # Longitud del tubo (m)
x_plot = np.linspace(0, L, 200)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Tubo abierto (Nodos de presión en x=0 y x=L): Seno(n*pi*x/L)
for n in [1, 2, 3]:
    y_open = np.sin(n * np.pi * x_plot / L)
    ax1.plot(x_plot, y_open, linewidth=2, label=f'Armónico {n} ($f_{n} = {n} f_1$)')
ax1.set_title('Tubo Abierto: Perfiles de Presión Acústica $\\Delta P(x)$', fontsize=12, fontweight='bold')
ax1.set_xlabel('Posición en el tubo X (m)', fontsize=11)
ax1.set_ylabel('Amplitud de Presión Normalizada', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True)

# Tubo cerrado (Antinodo en x=0, Nodo en x=L): Coseno((2n-1)*pi*x/2L)
for n in [1, 2, 3]:
    m = 2*n - 1
    y_closed = np.cos(m * np.pi * x_plot / (2 * L))
    ax2.plot(x_plot, y_closed, linewidth=2, label=f'Armónico {m} ($f_{{{m}}} = {m} f_1$)')
ax2.set_title('Tubo Cerrado: Perfiles de Presión Acústica $\\Delta P(x)$', fontsize=12, fontweight='bold')
ax2.set_xlabel('Posición en el tubo X (m)', fontsize=11)
ax2.grid(True, linestyle=':', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()

# 2. Cálculo matemático de las frecuencias de la escala temperada a partir de La 440
k_interval = 2**(1/12)
notes_names = ['La (440)', 'La# / Sib', 'Si', 'Do', 'Do# / Reb', 'Re', 'Re# / Mib', 'Mi', 'Fa', 'Fa# / Solb', 'Sol', 'Sol# / Lab', 'La (880)']
frequencies = [440 * (k_interval**i) for i in range(13)]

print('Frecuencias calculadas de la escala temperada (Octava 440Hz a 880Hz):\n')
print(f'{chr(124)} {"Nota":^15} {chr(124)} {"Frecuencia calculada (Hz)":^26} {chr(124)} {"Valor PDF (Hz)":^18} {chr(124)}')
print('-' * 67)
pdf_vals = [440.00, 466.16, 493.88, 523.25, 554.37, 587.33, 622.25, 659.26, 698.46, 739.99, 783.99, 830.61, 880.00]
for name, freq, pdf_v in zip(notes_names, frequencies, pdf_vals):
    print(f'{chr(124)} {name:<15} {chr(124)} {freq:^26.2f} {chr(124)} {pdf_v:^18.2f} {chr(124)}')
```

```python
# Simulación de síntesis armónica de Fourier para demostrar el Timbre

t_timbre = np.linspace(0, 0.01, 1000)
f0 = 220.0  # Frecuencia fundamental (La 220 Hz)
omega0 = 2 * np.pi * f0

# Generar ondas con diferentes componentes armónicos
# 1. Tono Puro (Solo frecuencia fundamental)
sine_pure = np.sin(omega0 * t_timbre)

# 2. Timbre tipo Onda Triangular (Armónicos impares con decaimiento 1/n^2)
triangle_wave = np.zeros_like(t_timbre)
for n in range(1, 6):
    harmonic = 2*n - 1
    amplitude = (-1)**(n-1) / (harmonic**2)
    triangle_wave += amplitude * np.sin(harmonic * omega0 * t_timbre)
    
# 3. Timbre tipo Onda Cuadrada (Armónicos impares con decaimiento 1/n)
square_wave = np.zeros_like(t_timbre)
for n in range(1, 6):
    harmonic = 2*n - 1
    amplitude = 1.0 / harmonic
    square_wave += amplitude * np.sin(harmonic * omega0 * t_timbre)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)
ax1.plot(t_timbre * 1000, sine_pure, 'b-', label='Frecuencia Fundamental Única (Flauta/Tono Puro)')
ax1.set_title('Diferentes Timbres Acústicos para una Frecuencia de 220 Hz', fontsize=12, fontweight='bold')
ax1.set_ylabel('Amplitud', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(loc='upper right')

ax2.plot(t_timbre * 1000, triangle_wave, 'g-', label='Armónicos $1/n^2$ (Clarinete/Onda Triangular)')
ax2.set_ylabel('Amplitud', fontsize=11)
ax2.grid(True, linestyle=':', alpha=0.5)
ax2.legend(loc='upper right')

ax3.plot(t_timbre * 1000, square_wave, 'r-', label='Armónicos $1/n$ (Oboe/Onda Cuadrada)')
ax3.set_xlabel('Tiempo (ms)', fontsize=11)
ax3.set_ylabel('Amplitud', fontsize=11)
ax3.grid(True, linestyle=':', alpha=0.5)
ax3.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

## 6. Efecto Doppler y Ondas de Choque

El **Efecto Doppler** describe el cambio en la frecuencia observada de una onda debido al movimiento relativo entre la fuente emisora y el receptor.

### 6.1 Formulación Matemática Clásica
En el caso clásico (sonido propagándose en el aire a velocidad mucho menor que la velocidad de la luz, $v \ll c$), las velocidades del emisor ($v_s$) y del receptor ($v_o$) se miden de forma absoluta respecto al medio (el aire). La frecuencia percibida por el observador $f_o$ en términos de la emitida por la fuente $f_f$ es:
$$f_o = f_f \left( \frac{v \pm v_o}{v \mp v_s} \right)$$
Donde $v$ es la velocidad del sonido en el medio. El criterio de signos sigue las siguientes reglas físicas:
* **Numerador ($v \pm v_o$):** Se utiliza el signo **$+$** si el observador se desplaza **hacia** la fuente (percibe frentes de onda con mayor rapidez, aumentando la frecuencia). Se utiliza el signo **$-$** si el observador se **aleja** de la fuente.
* **Denominador ($v \mp v_s$):** Se utiliza el signo **$-$** si la fuente se desplaza **hacia** el observador (acorta la distancia entre frentes de onda sucesivos, aumentando la frecuencia). Se utiliza el signo **$+$** si la fuente se **aleja** del observador.

### 6.2 Ondas de Choque y Número de Mach
Si la fuente emisora se mueve a una velocidad igual o mayor que la velocidad de propagación de las ondas en el medio ($v_s \geq v$), la ecuación clásica predice una frecuencia infinita (singularidad matemática). Físicamente, esto significa que la fuente avanza a la par o por delante de los frentes de onda circulares que ella misma emite.

El **Número Mach ($M$)** es la razón adimensional entre la velocidad de la fuente y la velocidad del sonido en el medio:
$$M = \frac{v_s}{v}$$
1. **Régimen Subsónico ($M < 1$):** Los frentes de onda se comprimen frente a la fuente pero viajan por delante de ella.
2. **Régimen Sónico ($M = 1$):** La fuente viaja exactamente a la velocidad del sonido, apilando todos los frentes de onda en una barrera de presión única frente a ella (barrera del sonido).
3. **Régimen Supersónico ($M > 1$):** La fuente se adelanta a las perturbaciones. Las ondas esféricas individuales se superponen de manera constructiva a lo largo de una envolvente cónica común. Esta envolvente es una **onda de choque** (envolvente cónica de alta presión).

El ángulo semi-vertical del cono de Mach ($\phi$) está dado geométricamente por la relación entre el camino recorrido por el sonido ($v \cdot t$) y el recorrido por la fuente ($v_s \cdot t$):
$$\sin\phi = \frac{v \cdot t}{v_s \cdot t} = \frac{v}{v_s} = \frac{1}{M}$$
El paso de esta onda de choque es escuchado por un observador terrestre estático como un chasquido o explosión muy fuerte llamado **estampido sónico** (sonic boom).


```python
# Simulación de los frentes de onda Doppler y formación del Cono de Mach

def simular_doppler(mach_number, fig_title):
    fig, ax = plt.subplots(figsize=(8, 8))
    v_wave = 1.0  # Velocidad de onda de referencia
    v_source = mach_number * v_wave  # Velocidad del emisor
    
    # Instantes de emisión de ondas del pasado (t_emision = 0, 1, 2, ..., 12)
    t_current = 14.0
    t_emissions = np.linspace(0, 12, 13)
    
    # Dibujar la trayectoria de la fuente
    x_source_track = v_source * np.linspace(0, t_current, 100)
    ax.plot(x_source_track, np.zeros_like(x_source_track), 'k--', alpha=0.5)
    
    # Dibujar frentes de onda
    for t_em in t_emissions:
        x_center = v_source * t_em
        radius = v_wave * (t_current - t_em)
        # Círculo de frente de onda
        circle = plt.Circle((x_center, 0), radius, fill=False, color='royalblue', linewidth=1.5, alpha=0.8)
        ax.add_patch(circle)
        # Punto de emisión
        ax.scatter(x_center, 0, color='gray', s=10, zorder=2)
        
    # Posición actual de la fuente
    x_source_curr = v_source * t_current
    ax.scatter(x_source_curr, 0, color='red', s=80, zorder=5, label='Fuente Emisora')
    
    # Dibujar cono de Mach si es supersónico
    if mach_number > 1.0:
        mach_angle = np.arcsin(1.0 / mach_number)
        # Línea tangente superior e inferior
        x_cone = np.linspace(0, x_source_curr, 100)
        y_cone_up = (x_source_curr - x_cone) * np.tan(mach_angle)
        y_cone_down = -y_cone_up
        ax.plot(x_cone, y_cone_up, 'r-', linewidth=2.5, label='Onda de Choque (Frente Cónico)')
        ax.plot(x_cone, y_cone_down, 'r-', linewidth=2.5)
        
        # Dibujar arco para mostrar el ángulo Mach
        arc_x = x_source_curr - 2.5 * np.cos(np.linspace(np.pi - mach_angle, np.pi, 50))
        arc_y = 2.5 * np.sin(np.linspace(np.pi - mach_angle, np.pi, 50))
        ax.plot(arc_x, arc_y, 'g-', linewidth=2)
        ax.text(x_source_curr - 4.5, 0.8, f'$\\phi = {np.degrees(mach_angle):.1f}^\\circ$', fontsize=11, color='green', fontweight='bold')
        
    ax.set_title(fig_title, fontsize=13, fontweight='bold')
    ax.set_xlabel('Posición X', fontsize=11)
    ax.set_ylabel('Posición Y', fontsize=11)
    ax.set_aspect('equal')
    ax.grid(True, linestyle=':', alpha=0.5)
    ax.legend(frameon=True, loc='upper left')
    
    # Ajustar límites automáticamente
    limit_val = x_source_curr + 3.0
    ax.set_xlim(-1, limit_val)
    ax.set_ylim(-limit_val/1.5, limit_val/1.5)
    plt.show()

# 1. Caso Subsónico (M = 0.6)
simular_doppler(0.6, 'Efecto Doppler Subsónico ($M = 0.6 < 1$)')

# 2. Caso Supersónico (M = 1.6)
simular_doppler(1.6, 'Frente de Onda Supersónico y Cono de Mach ($M = 1.6 > 1$)')
```

## 7. Verificación de Resultados Numéricos

En esta sección de verificación validamos que las funciones desarrolladas de forma manual en el cuaderno se ajustan con precisión a las relaciones físicas teóricas estándar de la termodinámica y acústica clásica.


```python
# Celda de Verificación Numérica
# Comprobación del cálculo Doppler frente a una validación analítica directa

def calcular_doppler_frec(f_f, v, v_o, v_s, obs_hacia_fuente=True, fuente_hacia_obs=True):
    # Aplicar el criterio de signos del formalismo general
    num_sign = 1.0 if obs_hacia_fuente else -1.0
    den_sign = -1.0 if fuente_hacia_obs else 1.0
    
    numerator = v + num_sign * v_o
    denominator = v + den_sign * v_s
    
    return f_f * (numerator / denominator)

# Escenario de prueba: Fuente emitiendo a 440 Hz (La), moviéndose a 30 m/s hacia el observador.
# El observador se mueve a 10 m/s hacia la fuente. Velocidad del sonido = 343 m/s.
f_source = 440.0
v_sound_val = 343.0
v_obs = 10.0
v_src = 30.0

# Cálculo numérico manual
f_observed_calc = calcular_doppler_frec(f_source, v_sound_val, v_obs, v_src,
                                        obs_hacia_fuente=True, fuente_hacia_obs=True)

# Validación matemática analítica independiente: f_o = 440 * (343 + 10) / (343 - 30)
f_observed_analytic = f_source * (v_sound_val + v_obs) / (v_sound_val - v_src)

# Comparación numérica usando límites de tolerancia de punto flotante
assert np.isclose(f_observed_calc, f_observed_analytic, atol=1e-5)

print('--- VALIDACIÓN DE PARÁMETROS ACÚSTICOS ---')
print(f'Frecuencia de la fuente: {f_source} Hz')
print(f'Frecuencia observada calculada: {f_observed_calc:.4f} Hz')
print(f'Frecuencia observada analítica: {f_observed_analytic:.4f} Hz')
print('Validación Doppler: ¡EXITOSA! (Los resultados coinciden exactamente)')
```

## 1. Diferencia de Fase y Coherencia

Las ondas electromagnéticas son oscilaciones de los campos eléctrico y magnético que se propagan en el espacio-tiempo. Cuando dos ondas de la misma frecuencia se encuentran en un punto del espacio, sus amplitudes se suman vectorialmente siguiendo el **principio de superposición**.

### 1.1 Coherencia Espacial y Temporal
Para que el patrón resultante sea observable y no promedie a una intensidad constante a lo largo del tiempo, es requisito indispensable que exista **coherencia**:
* **Coherencia temporal:** Mide la correlación de la fase de la onda en un mismo punto del espacio en diferentes instantes de tiempo. Está relacionada con la monocromaticidad de la fuente. La longitud de coherencia $L_c$ es la distancia máxima sobre la cual la onda mantiene una fase predecible:
  $$L_c = c \tau_c \approx \frac{\lambda^2}{\Delta \lambda}$$
  donde $\tau_c$ es el tiempo de coherencia y $\Delta \lambda$ es el ancho de banda espectral de la fuente. Un láser de gas tiene una gran longitud de coherencia (metros a kilómetros), mientras que la luz blanca tiene una longitud de coherencia muy corta (micrómetros).
* **Coherencia espacial:** Mide la correlación de la fase entre dos puntos diferentes del frente de onda en un mismo instante de tiempo. Es fundamental para poder obtener interferencia a partir de frentes de onda espacialmente extendidos.

### 1.2 Diferencia de Fase y Camino Óptico
Consideremos dos ondas armónicas coherentes con la misma frecuencia angular $\omega$ y amplitud $A$ que viajan distancias $x_1$ y $x_2$ desde sus fuentes al punto de observación $P$:
$$F_1 = A \sin(\omega t - k x_1)$$
$$F_2 = A \sin(\omega t - k x_2)$$
Donde $k = 2\pi / \lambda$ es el número de onda. La suma de ambas ondas es:
$$F_T = F_1 + F_2 = 2A \cos\left( \frac{k(x_2 - x_1)}{2} \right) \sin\left( \omega t - k\frac{x_1 + x_2}{2} \right)$$
La amplitud resultante es $A_T = 2A \cos\left( \frac{\Delta \phi}{2} \right)$, donde el desfase $\Delta \phi$ es:
$$\Delta \phi = k (x_2 - x_1) = \frac{2\pi}{\lambda} \Delta x$$
Donde $\Delta x = x_2 - x_1$ es la **diferencia de camino geométrico**. Si las ondas se propagan en un medio con índice de refracción $n$, definimos la **diferencia de camino óptico (OPD)** como $n \Delta x$.

* **Interferencia Constructiva (en fase):** Ocurre cuando el desfase es un múltiplo entero de $2\pi$:
  $$\Delta \phi = 2n\pi \implies \Delta x = n\lambda, \quad n \in \mathbb{Z}$$
  En este caso, la amplitud es máxima: $A_T = \pm 2A$, y la intensidad es cuatro veces la de una sola onda ($I \propto A_T^2 = 4A^2$).
* **Interferencia Destructiva (en contrafase):** Ocurre cuando el desfase es un múltiplo impar de $\pi$:
  $$\Delta \phi = (2n + 1)\pi \implies \Delta x = \left(n + \frac{1}{2}\right)\lambda, \quad n \in \mathbb{Z}$$
  La amplitud resultante es cero: $A_T = 0$, dando lugar a una franja oscura.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar el estilo de graficación científica
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Tiempo para la simulación
t = np.linspace(0, 4 * np.pi, 1000)
y1 = np.sin(t)
y2_constructive = np.sin(t)          # En fase (desfase = 0)
y2_destructive = np.sin(t + np.pi)    # En contrafase (desfase = pi)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 1. Interferencia Constructiva
ax1.plot(t, y1, 'b--', linewidth=1.5, label='Onda 1 ($y_1$)')
ax1.plot(t, y2_constructive, 'g:', linewidth=1.5, label='Onda 2 ($y_2$, en fase)')
ax1.plot(t, y1 + y2_constructive, 'r-', linewidth=2.5, label='Superposición ($y_1 + y_2$)')
ax1.set_title('Superposición: Interferencia Constructiva ($\\Delta\\phi = 0$)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Tiempo $t$ (rad)', fontsize=11)
ax1.set_ylabel('Amplitud', fontsize=11)
ax1.set_xlim(0, 4 * np.pi)
ax1.set_ylim(-2.2, 2.2)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True, loc='upper right')

# 2. Interferencia Destructiva
ax2.plot(t, y1, 'b--', linewidth=1.5, label='Onda 1 ($y_1$)')
ax2.plot(t, y2_destructive, 'g:', linewidth=1.5, label='Onda 2 ($y_2$, en contrafase)')
ax2.plot(t, y1 + y2_destructive, 'r-', linewidth=2.5, label='Superposición ($y_1 + y_2$)')
ax2.set_title('Superposición: Interferencia Destructiva ($\\Delta\\phi = \\pi$)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Tiempo $t$ (rad)', fontsize=11)
ax2.set_ylabel('Amplitud', fontsize=11)
ax2.set_xlim(0, 4 * np.pi)
ax2.set_ylim(-2.2, 2.2)
ax2.grid(True, linestyle=':', alpha=0.5)
ax2.legend(frameon=True, loc='upper right')

plt.tight_layout()
plt.show()
```

## 2. Interferencia en Películas Delgadas

La interferencia en películas delgadas se produce cuando la luz incide sobre una capa delgada de material transparente (como una pompa de jabón o una película de aceite sobre agua). El rayo incidente se refleja parcialmente en la superficie superior de la película y parcialmente se transmite. El rayo transmitido recorre el espesor de la película y se refleja en la superficie inferior, regresando e interfiriendo con la primera reflexión.

### 2.1 Saltos de Fase por Reflexión
Cuando una onda electromagnética que viaja en un medio con índice de refracción $n_1$ se refleja en la superficie de un medio con índice de refracción $n_2$:
* Si $n_2 > n_1$: La onda reflejada experimenta un cambio de fase de $\pi$ radianes (equivalente a media longitud de onda $\lambda/2$).
* Si $n_2 < n_1$: La onda reflejada no experimenta cambio de fase.

### 2.2 Ecuaciones de Interferencia (Película Plana de Espesor $t$)
Para una película de espesor $t$ e índice de refracción $n_2$, rodeada por medios de índices $n_1$ (superior) y $n_3$ (inferior), la diferencia de camino óptico para un ángulo de refracción $\theta_2$ en la película es:
$$\text{OPD} = 2 n_2 t \cos\theta_2$$
Debemos considerar el efecto combinado de la diferencia de camino óptico y los saltos de fase en ambas interfaces.

---

> [!WARNING]
> ### 🚨 Fe de Erratas: Corrección sobre el Sistema Aceite sobre Agua en la Literatura Académica
> En el PDF original del temario (pág. 8), se presenta el caso clásico de una película de aceite ($n_{\text{aceite}} \approx 1.5$) flotando sobre agua ($n_{\text{agua}} \approx 1.33$), estando en contacto con el aire ($n_{\text{aire}} = 1.0$). El texto original afirma incorrectamente que:
> - $2n_{\text{aceite}} t \cos\theta_2 = (m + 1/2)\lambda$ es la condición de interferencia **destructiva**.
> - $2n_{\text{aceite}} t \cos\theta_2 = m\lambda$ es la condición de interferencia **constructiva**.
> 
> **La derivación física correcta es la siguiente:**
> 1. **Interfaz Aire-Aceite ($n_{\text{aire}}=1.0 \rightarrow n_{\text{aceite}}=1.5$):** Como $n_{\text{aceite}} > n_{\text{aire}}$, el rayo reflejado en la superficie superior sufre un **salto de fase de $\pi$**.
> 2. **Interfaz Aceite-Agua ($n_{\text{aceite}}=1.5 \rightarrow n_{\text{agua}}=1.33$):** Como $n_{\text{agua}} < n_{\text{aceite}}$, el rayo transmitido se refleja en un medio de menor índice, por lo que **no sufre salto de fase** ($0$ radianes).
> 
> Debido a que solo una de las reflexiones sufre un cambio de fase de $\pi$, las dos ondas reflejadas parten con un desfase intrínseco de $\pi$. Para que interfieran **constructivamente** (en fase), la diferencia de camino óptico debe introducir un desfase adicional de $\pi$ (o un múltiplo impar de $\pi$), lo que equivale a media longitud de onda:
> $$\text{OPD} = 2 n_{\text{aceite}} t \cos\theta_2 = \left(m + \frac{1}{2}\right)\lambda, \quad m \in \{0, 1, 2, \dots\} \quad \text{(Constructiva)}$$
> Y para que interfieran **destructivamente** (en contrafase y se cancelen), la diferencia de camino óptico debe ser un número entero de longitudes de onda:
> $$\text{OPD} = 2 n_{\text{aceite}} t \cos\theta_2 = m\lambda, \quad m \in \{1, 2, 3, \dots\} \quad \text{(Destructiva)}$$
> 
> *Nota: Si la película fuera de agua ($n=1.33$) sobre vidrio ($n=1.5$), ambas interfaces experimentarían un salto de fase de $\pi$ ($n_{\text{aire}} < n_{\text{agua}} < n_{\text{vidrio}}$), cancelándose el efecto relativo del salto de fase y restableciéndose las condiciones estándar ($2nt\cos\theta = m\lambda$ para constructiva).*

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros físicos del sistema aire-aceite-agua
n_air = 1.0
n_oil = 1.5
n_water = 1.33

# Coeficientes de reflexión de Fresnel en incidencia normal
r1 = (n_air - n_oil) / (n_air + n_oil)        # -0.2 (salto de fase de pi)
r2 = (n_oil - n_water) / (n_oil + n_water)    # ~0.0601 (sin salto de fase)

# Rango de longitudes de onda en el espectro visible (nm)
wavelengths = np.linspace(400, 750, 1000)

# Espesores de la película de aceite a simular (nm)
espesores = [150, 250, 350]
colors = ['purple', 'teal', 'darkorange']

plt.figure(figsize=(11, 6))

# Dibujar el fondo del espectro visible para referencia física
plt.axvspan(400, 450, color='violet', alpha=0.15)
plt.axvspan(450, 495, color='blue', alpha=0.15)
plt.axvspan(495, 570, color='green', alpha=0.15)
plt.axvspan(570, 590, color='yellow', alpha=0.15)
plt.axvspan(590, 620, color='orange', alpha=0.15)
plt.axvspan(620, 750, color='red', alpha=0.15)

# Calcular y graficar la reflectancia para cada espesor
for t, col in zip(espesores, colors):
    # Diferencia de fase debida al camino óptico en incidencia normal (cos(theta_2) = 1)
    # Reflectancia de potencia usando la aproximación de doble haz R = |r1 + r2 * exp(-i*delta_path)|^2
    # R = r1^2 + r2^2 + 2*r1*r2*cos(4 * pi * n_oil * t / lambda)
    # Nota: r1 es negativo (-0.2), lo que representa matemáticamente el salto de fase en la superficie superior
    R = r1**2 + r2**2 + 2 * r1 * r2 * np.cos(4 * np.pi * n_oil * t / wavelengths)
    plt.plot(wavelengths, R, color=col, linewidth=2.5, label=f'Espesor de aceite $t = {t}$ nm')

plt.title('Espectro de Reflectancia de una Película de Aceite ($n=1.5$) sobre Agua ($n=1.33$)', fontsize=13, fontweight='bold')
plt.xlabel('Longitud de onda $\\lambda$ (nm)', fontsize=11)
plt.ylabel('Reflectancia de Potencia $R$', fontsize=11)
plt.xlim(400, 750)
plt.ylim(0, 0.08)
plt.grid(True, linestyle=':', alpha=0.4)
plt.legend(frameon=True, loc='upper right')

# Mostrar algunas anotaciones de máximos y mínimos para el espesor de 250 nm
# Constructivo teórico para m=1: lam = 2 * 1.5 * 250 / 1.5 = 500 nm (Verde)
# Destructivo teórico para m=1: lam = 2 * 1.5 * 250 / 1 = 750 nm (Rojo lejano)
plt.annotate('Máximo Constructivo (500 nm)\npara $t=250$ nm ($m=1$)', xy=(500, 0.067), xytext=(520, 0.075),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6), fontsize=9)
plt.annotate('Mínimo Destructivo (750 nm)\npara $t=250$ nm ($m=1$)', xy=(750, 0.019), xytext=(630, 0.005),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6), fontsize=9)

plt.tight_layout()
plt.show()
```

## 3. Interferencia en Dos Rendijas: Experimento de Young

En 1801, Thomas Young demostró la naturaleza ondulatoria de la luz mediante el experimento de la doble rendija. Al iluminar dos rendijas estrechas paralelas separadas por una distancia $d$ con luz coherente y monocromática de longitud de onda $\lambda$, se observa en una pantalla distante a una distancia $L$ ($L \gg d$) un patrón alternante de franjas brillantes y oscuras.

### 3.1 Diferencia de Camino en Aproximación de Campo Lejano
Para un punto $P$ en la pantalla situado a una distancia $y$ del eje óptico (centro de la pantalla), los dos rayos que emergen de las rendijas viajan caminos con una diferencia $\Delta x = d \sin\theta$.
En la aproximación paraxial (ángulo $\theta$ muy pequeño), se cumple que $\sin\theta \approx \tan\theta = y / L$. Por lo tanto:
$$\Delta x \approx d \frac{y}{L}$$

* **Franjas Brillantes (Máximos - Interferencia Constructiva):**
  $$d \sin\theta = n\lambda \implies y_{\text{brillante}} = \frac{n \lambda L}{d}, \quad n \in \mathbb{Z}$$
* **Franjas Oscuras (Mínimos - Interferencia Destructiva):**
  $$d \sin\theta = \left(n + \frac{1}{2}\right)\lambda \implies y_{\text{oscura}} = \frac{\left(n + \frac{1}{2}\right) \lambda L}{d}, \quad n \in \mathbb{Z}$$

El espaciamiento entre dos franjas brillantes consecutivas (ancho de franja $w$) es constante en la aproximación paraxial:
$$w = y_{n+1} - y_n = \frac{\lambda L}{d}$$

### 3.2 Distribución de Intensidad
Sumando las fases de las dos fuentes puntuales coherentes, la intensidad resultante $I$ en la pantalla como función del ángulo $\theta$ viene dada por:
$$I(\theta) = I_0 \cos^2\left( \frac{\pi d \sin\theta}{\lambda} \right) \approx I_0 \cos^2\left( \frac{\pi d y}{\lambda L} \right)$$
Donde $I_0$ es la intensidad en el centro (máximo principal).


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros experimentales
d = 0.25e-3   # Separación entre rendijas (0.25 mm)
L = 1.8       # Distancia a la pantalla (1.8 m)
lambda_red = 633e-9   # Láser Rojo He-Ne (633 nm)
lambda_blue = 450e-9  # Láser Azul de Diodo (450 nm)

# Rango en la pantalla (y de -20 mm a 20 mm)
y = np.linspace(-20e-3, 20e-3, 1000)

# Intensidades normalizadas
I_red = np.cos(np.pi * d * y / (lambda_red * L))**2
I_blue = np.cos(np.pi * d * y / (lambda_blue * L))**2

# Ancho de franja teórico
w_red = lambda_red * L / d
w_blue = lambda_blue * L / d

plt.figure(figsize=(11, 5.5))
plt.plot(y * 1000, I_red, 'r-', linewidth=2.0, label=f'Láser Rojo ($\\lambda=633$ nm), $w \\approx {w_red*1e3:.2f}$ mm')
plt.plot(y * 1000, I_blue, 'b-', linewidth=1.5, alpha=0.7, label=f'Láser Azul ($\\lambda=450$ nm), $w \\approx {w_blue*1e3:.2f}$ mm')

plt.title('Distribución de Intensidad en la Pantalla (Experimento de Young)', fontsize=13, fontweight='bold')
plt.xlabel('Posición en la pantalla $y$ (mm)', fontsize=11)
plt.ylabel('Intensidad Normalizada $I/I_0$', fontsize=11)
plt.xlim(-20, 20)
plt.ylim(-0.05, 1.05)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

## 4. Difracción por una Rendija y Patrón Combinado

Cuando la luz pasa a través de una sola rendija de ancho finito $a$, cada punto del frente de onda dentro de la rendija actúa como una fuente secundaria de ondas (principio de Huygens-Fresnel). Estas fuentes secundarias interfieren entre sí, dando lugar al patrón de **difracción de Fraunhofer**.

### 4.1 Difracción de Fraunhofer por Rendija Única
La intensidad en la pantalla en función del ángulo $\theta$ es:
$$I_D(\theta) = I_0 \operatorname{sinc}^2\left( \frac{\pi a \sin\theta}{\lambda} \right)$$
Donde $\operatorname{sinc}(u) = \frac{\sin(u)}{u}$ es la función seno cardinal no normalizada.
* **Mínimos de Difracción:** Ocurren cuando el numerador es cero pero el argumento no es nulo:
  $$a \sin\theta_m = m\lambda, \quad m \in \{\pm 1, \pm 2, \dots\}$$
  El máximo central tiene un ancho angular entre los primeros mínimos laterales ($m = \pm 1$) igual a $2 \theta_1 \approx 2\lambda/a$.

### 4.2 Patrón Combinado para Dos Rendijas de Ancho Finito
En un experimento real con dos rendijas de separación $d$ y ancho individual $a$, el patrón observado es el producto de la interferencia de dos rendijas y la difracción de cada rendija individual. La difracción actúa como una "envolvente" que modula las franjas de interferencia rápidas:
$$I(\theta) = I_0 \cos^2\left( \frac{\pi d \sin\theta}{\lambda} \right) \operatorname{sinc}^2\left( \frac{\pi a \sin\theta}{\lambda} \right)$$
Donde la condición de existencia física requiere que la separación de rendijas sea mayor que el ancho de la rendija ($d > a$).


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
a = 0.05e-3     # Ancho de rendija (0.05 mm)
d = 0.22e-3     # Separación entre rendijas (0.22 mm)
L = 1.5         # Distancia a la pantalla (1.5 m)
lam = 633e-9    # Longitud de onda del láser (633 nm)

# Posición en pantalla (y de -40 mm a 40 mm)
y = np.linspace(-40e-3, 40e-3, 1000)

# Cálculos teóricos
alpha = np.pi * d * y / (lam * L)
beta = np.pi * a * y / (lam * L)

# numpy.sinc(x) calcula sin(pi*x)/(pi*x), por lo que debemos dividir el argumento beta por pi
I_diff = np.sinc(a * y / (lam * L))**2
I_interf = np.cos(alpha)**2
I_comb = I_interf * I_diff

plt.figure(figsize=(11, 5.5))
plt.plot(y * 1000, I_comb, 'r-', linewidth=2.0, label='Patrón Combinado (Interferencia + Difracción)')
plt.plot(y * 1000, I_diff, 'k--', linewidth=1.5, alpha=0.7, label='Envolvente de Difracción (Rendija Única)')
plt.plot(y * 1000, -I_diff, 'k--', linewidth=1.5, alpha=0.7)  # Envolvente inferior

plt.title('Interferencia de Doble Rendija Modulada por Difracción (Ancho de Rendija $a = 0.05$ mm)', fontsize=13, fontweight='bold')
plt.xlabel('Posición en la pantalla $y$ (mm)', fontsize=11)
plt.ylabel('Intensidad Normalizada $I/I_0$', fontsize=11)
plt.xlim(-40, 40)
plt.ylim(-0.15, 1.05)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

## 5. Interferómetro de Michelson

El interferómetro de Michelson, diseñado por Albert A. Michelson a finales del siglo XIX, es un instrumento óptico de alta precisión que divide un haz de luz en dos caminos perpendiculares utilizando un divisor de haz (beam splitter). Los dos haces se reflejan en espejos planos y vuelven a combinarse en el divisor de haz antes de llegar a un detector o pantalla.

### 5.1 Principio de Funcionamiento y Franjas de Interferencia
Sean $L_1$ y $L_2$ las longitudes de los dos brazos del interferómetro. Un haz que recorre el brazo 1 recorre una distancia de ida y vuelta de $2L_1$, mientras que en el brazo 2 recorre $2L_2$. La diferencia de camino óptico en el centro de la pantalla es:
$$\text{OPD} = 2(L_1 - L_2) = 2\Delta L$$

Si la luz incide con un cierto ángulo $\theta$ respecto al eje de los espejos (por ejemplo, mediante un haz ligeramente divergente o espejos no perfectamente alineados), la diferencia de camino óptico depende del ángulo:
$$\text{OPD} = 2\Delta L \cos\theta$$

La intensidad en la pantalla para luz monocromática es:
$$I(\theta) = I_0 \cos^2\left( \frac{2\pi \Delta L \cos\theta}{\lambda} \right)$$

* **Franjas circulares (anillos concéntricos):** Se forman cuando los espejos son perfectamente perpendiculares y el haz de luz es divergente. Los puntos de igual inclinación $\theta$ forman círculos concéntricos en la pantalla.
* **Movimiento del espejo móvil:** Si uno de los espejos se desplaza una distancia $d_m$, la diferencia de camino óptico cambia en $2d_m$. Cada vez que el espejo se desplaza una distancia de $\lambda/2$, el patrón de interferencia cambia por una franja completa en el centro de la pantalla. El número de franjas $N$ que transitan por el centro de la pantalla al mover el espejo una distancia $d_m$ es:
  $$N = \frac{2 d_m}{\lambda}$$

### 5.2 Aplicaciones Científicas
1. **Medición de longitudes de onda:** Contando el número de franjas al mover el espejo móvil una distancia conocida con precisión micrométrica.
2. **Espectroscopía por Transformada de Fourier (FTIR):** Permite determinar la composición espectral de luz poli-cromática midiendo el interferograma y aplicando la Transformada de Fourier.
3. **Detección de Ondas Gravitacionales (LIGO):** Los brazos del interferómetro LIGO miden 4 km de longitud. Las ondas gravitacionales causan distorsiones en el espacio-tiempo que cambian imperceptiblemente la longitud relativa de los brazos ($\sim 10^{-19}$ m), lo cual altera la condición de interferencia constructiva/destructiva y se detecta ópticamente.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
lam = 633e-9    # Láser de Helio-Neon (633 nm)
Ls = 0.5        # Distancia a la pantalla (0.5 m)

# Brazo 1: Diferencia de caminos para la primera configuración (0.1 mm)
delta_L1 = 0.1e-3
# Brazo 2: Desplazamiento de una fracción de longitud de onda (0.1 mm + lam/4)
# Un desplazamiento del espejo de lam/4 causa un cambio de fase de pi en el centro (anillos se desplazan)
delta_L2 = 0.1e-3 + 0.25 * lam

# Grid espacial en la pantalla (de -8 mm a 8 mm)
size = 8e-3
x = np.linspace(-size, size, 400)
y = np.linspace(-size, size, 400)
X, Y = np.meshgrid(x, y)
r = np.sqrt(X**2 + Y**2)

# cos(theta) de los rayos divergentes respecto al eje
cos_theta = Ls / np.sqrt(Ls**2 + r**2)

# Intensidades en pantalla
I1 = np.cos(2 * np.pi * delta_L1 * cos_theta / lam)**2
I2 = np.cos(2 * np.pi * delta_L2 * cos_theta / lam)**2

# Graficación de los dos patrones uno al lado del otro
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 6))

im1 = ax1.imshow(I1, extent=[-8, 8, -8, 8], cmap='Reds', origin='lower')
ax1.set_title(f'Patrón de Anillos para $\\Delta L = {delta_L1*1e3:.4f}$ mm', fontsize=12, fontweight='bold')
ax1.set_xlabel('$x$ (mm)', fontsize=10)
ax1.set_ylabel('$y$ (mm)', fontsize=10)
ax1.grid(False)  # Quitar la cuadrícula para el patrón de imagen
fig.colorbar(im1, ax=ax1, label='Intensidad Relativa', shrink=0.7)

im2 = ax2.imshow(I2, extent=[-8, 8, -8, 8], cmap='Reds', origin='lower')
ax2.set_title(f'Patrón de Anillos para $\\Delta L = {delta_L2*1e3:.4f}$ mm (Desplazado $\\lambda/4$ del espejo)', fontsize=12, fontweight='bold')
ax2.set_xlabel('$x$ (mm)', fontsize=10)
ax2.set_ylabel('$y$ (mm)', fontsize=10)
ax2.grid(False)
fig.colorbar(im2, ax=ax2, label='Intensidad Relativa', shrink=0.7)

plt.suptitle('Patrón 2D de Interferencia del Interferómetro de Michelson', fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```

## 6. Validación Numérica y Verificación de la Física

Para consolidar el aprendizaje y garantizar la exactitud de los modelos implementados, ejecutaremos una celda de validación en Python. Utilizaremos aserciones (`assert`) para comprobar que:
1. Las condiciones de interferencia en películas de aceite sobre agua correspondan correctamente a máximos y mínimos con la corrección física de la errata (un solo salto de fase de $\pi$).
2. El espaciamiento entre franjas de Young ($w$) calculado a partir del perfil simulado coincida con la fórmula teórica $w = \lambda L / d$.
3. La relación entre desplazamiento del espejo del interferómetro de Michelson y el número de franjas transitorias sea exacta.


```python
# Validación de la física mediante aserciones numéricas

# 1. Validación de la película delgada de aceite sobre agua (n_oil = 1.5, n_water = 1.33)
# Para incidencia normal y un solo salto de fase de pi:
# Mínimos (destructiva) ocurren cuando 2 * n_oil * t = m * lam
# Máximos (constructiva) ocurren cuando 2 * n_oil * t = (m + 0.5) * lam
n_oil = 1.5
t_film = 250.0  # nm

# Para m = 1, la longitud de onda de interferencia destructiva teórica es:
# lam_dest = 2 * n_oil * t_film / 1 = 2 * 1.5 * 250 = 750 nm
lam_dest_teorica = 2 * n_oil * t_film / 1.0
print(f'Longitud de onda teórica para interferencia destructiva (m=1): {lam_dest_teorica} nm')

# Para m = 1, la longitud de onda de interferencia constructiva teórica es:
# lam_const = 2 * n_oil * t_film / (1 + 0.5) = 2 * 1.5 * 250 / 1.5 = 500 nm
lam_const_teorica = 2 * n_oil * t_film / 1.5
print(f'Longitud de onda teórica para interferencia constructiva (m=1): {lam_const_teorica} nm')

r1 = (1.0 - n_oil) / (1.0 + n_oil)  # -0.2 (salto de fase)
r2 = (n_oil - 1.33) / (n_oil + 1.33)  # ~0.0601 (sin salto)

def calc_R(lam):
    return r1**2 + r2**2 + 2 * r1 * r2 * np.cos(4 * np.pi * n_oil * t_film / lam)

R_dest = calc_R(lam_dest_teorica)
R_const = calc_R(lam_const_teorica)

print(f'Reflectancia en mínimo (destructiva, 750 nm): {R_dest:.6f}')
print(f'Reflectancia en máximo (constructiva, 500 nm): {R_const:.6f}')

# Aserciones
assert np.isclose(R_dest, (r1 + r2)**2), 'Error en el mínimo de interferencia en películas delgadas'
assert np.isclose(R_const, (r1 - r2)**2), 'Error en el máximo de interferencia en películas delgadas'
assert R_const > R_dest, 'La reflectancia constructiva debe ser mayor que la destructiva'

# 2. Validación de Young
d_slit = 0.25e-3  # m
L_dist = 1.8     # m
lam_laser = 633e-9 # m

w_teorico = lam_laser * L_dist / d_slit  # m
print(f'Ancho de franja de Young teórico: {w_teorico * 1e3:.4f} mm')

# A partir del perfil de intensidad simulado: I = cos^2(pi * d_slit * y / (lam_laser * L_dist))
# Los máximos ocurren en y = n * w_teorico. Por lo tanto, el espaciamiento entre el máximo central (y=0)
# y el primer máximo lateral (y = w_teorico) debe ser exactamente w_teorico.
y_puntos = np.array([0.0, w_teorico])
I_val = np.cos(np.pi * d_slit * y_puntos / (lam_laser * L_dist))**2
assert np.allclose(I_val, 1.0), 'Los máximos de Young no coinciden con la separación teórica'

# 3. Validación de Michelson
# Al desplazar el espejo d_m = 10 * lam_laser, debemos ver N = 20 franjas pasar por el centro.
d_m = 10 * lam_laser
N_franjas = 2 * d_m / lam_laser
print(f'Número de franjas teóricas para desplazamiento de espejo de {d_m*1e6:.3f} um: {N_franjas}')
assert np.isclose(N_franjas, 20.0), 'La relación del conteo de franjas de Michelson es incorrecta'

print('\n¡Todas las validaciones físicas y matemáticas se completaron con éxito!')
```

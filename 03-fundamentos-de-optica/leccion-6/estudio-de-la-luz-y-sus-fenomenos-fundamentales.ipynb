## 1. Naturaleza de la Luz y Dualidad Onda-Partícula

La naturaleza física de la luz ha sido uno de los debates más largos y profundos de la ciencia:
1. **Concepciones de la Antigüedad:** Empédocles consideraba que la luz era una sustancia física en movimiento emitida por los objetos. Platón la asociaba al conocimiento racional. Euclides formuló las bases de la propagación rectilínea y la reflexión en espejos. Aristóteles propuso que la luz era un movimiento en un medio transparente, análogo al sonido.
2. **Siglo XVII y XVIII (Newton vs. Huygens):** Isaac Newton propuso la **teoría corpuscular**, postulando que la luz consiste en una corriente de partículas rápidas (corpúsculos). Christian Huygens defendió la **teoría ondulatoria**, argumentando que la luz es una onda propagándose en el éter. La visión corpuscular dominó por el prestigio de Newton.
3. **Siglo XIX (Fresnel, Young y Maxwell):** Augustin-Jean Fresnel demostró que la luz es una onda transversal. Thomas Young realizó el experimento de la doble rendija en 1801, confirmando la naturaleza ondulatoria mediante la interferencia de luz. James Clerk Maxwell unificó la electricidad y el magnetismo, revelando que la luz es una onda electromagnética.
4. **Siglo XX (Einstein y de Broglie):** Albert Einstein propuso en 1905 que la luz se compone de cuantos de energía localizados llamados **fotones** para explicar el efecto fotoeléctrico. Louis de Broglie (1924) generalizó esta dualidad onda-partícula a todas las partículas elementales de la materia.

### 1.1 Relación de de Broglie
Cada partícula material con momento lineal $p = mv$ tiene asociada una longitud de onda de de Broglie $\lambda$ dada por:
$$\lambda = \frac{h}{p} = \frac{h}{mv}$$
Reordenando, el momento lineal de cualquier partícula (con o sin masa, como el fotón) se expresa como:
$$p = \frac{h}{\lambda}$$
Donde $h \approx 6.626 \times 10^{-34}\text{ J s}$ es la constante de Planck.

### 1.2 El Principio de Incertidumbre de Heisenberg
Si un electrón o un fotón se comporta como onda, su posición y su momento lineal no pueden determinarse simultáneamente con precisión absoluta. Werner Heisenberg (1927) formuló el **principio de incertidumbre**:
$$\Delta x \cdot \Delta p \ge \frac{\hbar}{2} \approx h$$
Donde $\Delta x$ es la incertidumbre en la posición y $\Delta p$ en el momento. Este principio impone un límite fundamental a la aproximación de la física clásica: a escalas atómicas, no podemos hablar de trayectorias exactas ni determinismo, sino de probabilidades de localización y estados ondulatorios de superposición.

### 1.3 Principio de Correspondencia de Bohr
Niels Bohr enunció que las leyes de la mecánica cuántica deben reducirse a las leyes de la física clásica en el límite de números cuánticos grandes o a escala macroscópica, actuando la física clásica como una excelente aproximación en el mundo cotidiano.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Parámetros del experimento de Young para electrones
wavelength = 0.05e-9  # Longitud de onda de de Broglie (m)
d_slits = 0.5e-6     # Distancia entre rendijas (m)
a_width = 0.08e-6     # Ancho de las rendijas (m)
D_distance = 1.0     # Distancia a la pantalla (m)

# Coordenada y vertical de la pantalla
y_screen = np.linspace(-0.6e-3, 0.6e-3, 1000)

# Densidad de probabilidad de impacto P(y)
beta = (np.pi * a_width * y_screen) / (wavelength * D_distance)
phi = (2 * np.pi / wavelength) * (d_slits * y_screen / D_distance)
P_y = (np.cos(phi / 2))**2 * (np.sinc(beta / np.pi))**2
P_y /= np.sum(P_y)  # Normalizar

# Generar impactos individuales de electrones (simulación cuántica de Monte Carlo)
num_electrons = 1500
electron_y = np.random.choice(y_screen, size=num_electrons, p=P_y)
electron_x = np.random.uniform(0.1, 0.9, num_electrons)  # Posición X en la franja detector

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 7), gridspec_kw={'height_ratios': [3, 1]})

# Panel 1: Patrón continuo de probabilidad (envolvente e interferencia)
ax1.plot(y_screen * 1e3, P_y * 1e3, color='royalblue', linewidth=2, label='Patrón Ondulatorio Teórico $P(y)$')
ax1.set_title('Dualidad Onda-Partícula: Interferencia Cuántica de Electrones Individuales', fontsize=13, fontweight='bold')
ax1.set_ylabel('Probabilidad Relativa', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True)

# Panel 2: Impactos de partículas individuales en el detector
ax2.scatter(electron_y * 1e3, electron_x, color='darkgreen', s=3, alpha=0.6, label='Electrones detectados')
ax2.set_xlabel('Posición en la pantalla Y (mm)', fontsize=11)
ax2.set_ylim(0, 1.0)
ax2.get_yaxis().set_visible(False)
ax2.legend(frameon=True, loc='upper right')

plt.tight_layout()
plt.show()
```

## 2. La Velocidad de la Luz

La velocidad de la luz ($c$) es una de las constantes más importantes del universo físico. Históricamente, determinar si era finita requirió de ingeniosos experimentos astronómicos y terrestres:

1. **Intento de Galileo:** Galileo intentó medir la velocidad usando linternas cubiertas y descubiertas por dos observadores en colinas distantes. El tiempo de reacción de los humanos superaba por millones de veces el tiempo de tránsito, concluyendo que la velocidad era "extraordinariamente rápida".
2. **Método de Ole Roemer (1676):** Estudió los eclipses de Ío, una de las lunas galileanas de Júpiter. Al comparar las predicciones del eclipse cuando la Tierra estaba en máximo acercamiento a Júpiter con los tiempos observados cuando la Tierra estaba en el punto más alejado, observó un retraso temporal acumulativo. Roemer dedujo que la luz tardaba $\sim 22\text{ minutos}$ en recorrer el diámetro de la órbita terrestre, estimando la velocidad en $220\,000\text{ km/s}$.
3. **Aberración de Bradley (1725):** James Bradley descubrió la aberración estelar: el cambio aparente de dirección de una estrella debido al movimiento orbital de la Tierra. Usando la analogía de la lluvia que cae verticalmente (donde debemos inclinar el paraguas al correr), Bradley calculó el ángulo de inclinación de la luz estelar $\theta$:
  $$\tan\theta = \frac{v_{\text{Tierra}}}{c}$$
  Esto arrojó un valor muy preciso de $c \approx 301\,000\text{ km/s}$.
4. **Predicción de Maxwell:** Demostró teóricamente que $c$ es constante e intrínseca al vacío:
  $$c = \frac{1}{\sqrt{\mu_0 \epsilon_0}} \approx 3 \times 10^8\text{ m/s}$$


```python
# Simulación de las órbitas de la Tierra e Io para ilustrar el retraso de Roemer
# El retardo en minutos depende de la distancia Tierra-Júpiter a lo largo del año

theta_earth = np.linspace(0, 2 * np.pi, 300)  # Ángulo de la Tierra en su órbita (1 año)
r_earth = 1.0  # Unidad Astronómica (UA)
r_jupiter = 5.2  # UA

# Para simplificar, suponemos a Júpiter estático en theta = 0 en la escala de un año
x_jupiter, y_jupiter = r_jupiter, 0.0
x_earth = r_earth * np.cos(theta_earth)
y_earth = r_earth * np.sin(theta_earth)

# Distancia Tierra-Júpiter
distance_EJ = np.sqrt((x_earth - x_jupiter)**2 + (y_earth - y_jupiter)**2)

# Retardo temporal en el viaje de la luz (d - d_min) / c
# 1 UA = 1.496e11 m, c = 2.998e8 m/s => 1 UA / c = 499 segundos = 8.31 minutos
c_in_UA_per_min = 1.0 / 8.31  # UA / min
d_min = r_jupiter - r_earth
delay_minutes = (distance_EJ - d_min) / c_in_UA_per_min

plt.figure(figsize=(10, 5))
plt.plot(np.degrees(theta_earth), delay_minutes, color='darkorange', linewidth=2.5, label='Retardo del Eclipse')
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.title('Simulación del Retardo de Roemer para los Eclipses de Ío', fontsize=13, fontweight='bold')
plt.xlabel('Ángulo orbital de la Tierra relativo a Júpiter (grados)', fontsize=11)
plt.ylabel('Retardo acumulado de la señal (minutos)', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.5)
plt.xlim(0, 360)
plt.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

## 3. Reflexión, Refracción y Dispersión de la Luz

Al cambiar de interfaz, la luz altera su comportamiento geométrico:
* **Reflexión:** Rebote de los frentes de onda en una frontera.
  - *Especular:* En superficies planas, con ángulo incidente igual al reflejado ($\theta_i = \theta_r$). La imagen se invierte en el eje normal.
  - *Difusa:* En superficies rugosas, donde los frentes se dispersan en múltiples direcciones debido a imperfecciones y centros de dispersión secundarios.
* **Refracción:** Desviación por cambio de velocidad al cambiar de medio elástico, descrita por la **ley de Snell**:
  $$n_1 \sin\theta_1 = n_2 \sin\theta_2$$
  Donde $n_i = c/v_i$ es el índice de refracción del medio.

### 3.1 Dispersión Cromática y la Ecuación de Cauchy
El índice de refracción de un material no es estrictamente constante; varía en función de la longitud de onda de la luz incidente $\lambda$. Augustin-Louis Cauchy dedujo una corrección empírica para esta dispersión cromática en medios dieléctricos transparentes:
$$n(\lambda) = A + \frac{B}{\lambda^2} + \frac{C}{\lambda^4} + \dots \approx A + \frac{B}{\lambda^2}$$
Donde $A$ y $B$ son coeficientes experimentales del material. Dado que $n$ es inversamente proporcional a $\lambda^2$, las longitudes de onda más cortas (como el azul o violeta) experimentan un índice de refracción mayor que las longitudes de onda largas (como el rojo). Esto explica la separación del espectro visible al atravesar un prisma y es el origen de la **aberración cromática** en lentes.


```python
# Simulación de refracción dispersiva en un prisma triangular equilátero de 60 grados

alpha = np.radians(60.0)  # Ángulo del prisma
theta_i1 = np.radians(45.0)  # Ángulo de incidencia en la primera cara

# Coeficientes de Cauchy para vidrio Flint
A_flint = 1.62
B_flint = 1.3e-14  # m^2

# Longitudes de onda a evaluar (m)
wavelengths_nm = {'Rojo': 650e-9, 'Verde': 532e-9, 'Azul': 450e-9}
colors_rgb = {'Rojo': 'red', 'Verde': 'green', 'Azul': 'blue'}

print('Separación angular en la salida del prisma Flint:\n')
for name, lam in wavelengths_nm.items():
    # Índice de refracción según la fórmula de Cauchy
    n_glass = A_flint + B_flint / lam**2
    
    # 1era Refracción (Aire -> Vidrio)
    sin_theta_r1 = np.sin(theta_i1) / n_glass
    theta_r1 = np.arcsin(sin_theta_r1)
    
    # Geometría del prisma: el ángulo de incidencia interno es theta_i2
    theta_i2 = alpha - theta_r1
    
    # 2da Refracción (Vidrio -> Aire)
    sin_theta_r2 = n_glass * np.sin(theta_i2)
    if sin_theta_r2 <= 1.0:
        theta_r2 = np.arcsin(sin_theta_r2)
        deviation = (theta_i1 - theta_r1) + (theta_r2 - theta_i2)
        print(f'* {name:<7} (n={n_glass:.4f}): Ángulo de emergencia = {np.degrees(theta_r2):.2f}° | Desviación = {np.degrees(deviation):.2f}°')
    else:
        print(f'* {name:<7} (n={n_glass:.4f}): Reflexión Interna Total en la segunda cara')
```

## 4. Reflexión Total Interna y Ángulo Crítico

Cuando la luz viaja desde un medio ópticamente más denso ($n_1$) a uno menos denso ($n_2$), con $n_1 > n_2$, los rayos refractados se desvían alejándose de la normal.
Al aumentar el ángulo de incidencia $\theta_1$, llega un punto donde el ángulo de refracción $\theta_2$ alcanza los $90^\circ$. En este límite, el rayo refractado viaja de forma paralela a la superficie de interfaz. Este ángulo de incidencia límite se denomina **ángulo crítico** ($\theta_c$):
$$n_1 \sin\theta_c = n_2 \sin(90^\circ) \implies \sin\theta_c = \frac{n_2}{n_1} \implies \theta_c = \arcsin\left( \frac{n_2}{n_1} \right)$$

Si el ángulo de incidencia supera el crítico ($\theta_1 > \theta_c$), no hay transmisión al segundo medio; el $100\%$ de la energía luminosa es reflejada de vuelta al medio original. Este fenómeno se conoce como **reflexión total interna** y es la base tecnológica de la fibra óptica.


```python
# Graficar rayos de luz incidiendo de agua (n=1.33) a aire (n=1.0)
n1_water = 1.33
n2_air = 1.0
theta_c_water = np.arcsin(n2_air / n1_water)

angles_deg = [15, 30, 45, 48.75, 60, 75]  # Ángulos a graficar

plt.figure(figsize=(10, 8))
# Dibujar interfaz
plt.axhline(0, color='cyan', linewidth=3, label='Interfaz Agua-Aire')
plt.fill_between([-2, 2], -2, 0, color='lightblue', alpha=0.4, label='Medio 1: Agua (n=1.33)')
plt.fill_between([-2, 2], 0, 2, color='whitesmoke', alpha=0.5, label='Medio 2: Aire (n=1.0)')
plt.axvline(0, color='black', linestyle=':', alpha=0.7, label='Normal a la superficie')

for ang in angles_deg:
    theta1 = np.radians(ang)
    # Rayo incidente (viaja de abajo hacia arriba)
    x_inc = [-np.sin(theta1), 0]
    y_inc = [-np.cos(theta1), 0]
    
    if theta1 < theta_c_water:
        # Rayo refractado en el aire
        sin_theta2 = (n1_water * np.sin(theta1)) / n2_air
        theta2 = np.arcsin(sin_theta2)
        x_ref = [0, np.sin(theta2)]
        y_ref = [0, np.cos(theta2)]
        plt.plot(x_inc, y_inc, 'b-', linewidth=1.5)
        plt.plot(x_ref, y_ref, 'g-', linewidth=2)
    else:
        # Reflexión total interna
        x_refl = [0, np.sin(theta1)]
        y_refl = [0, -np.cos(theta1)]
        plt.plot(x_inc, y_inc, 'b-', linewidth=1.5)
        plt.plot(x_refl, y_refl, 'r-', linewidth=2.5)

plt.title('Reflexión Total Interna y Ángulo Crítico $\\theta_c \\approx 48.7^\\circ$', fontsize=13, fontweight='bold')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='lower left')
plt.gca().set_aspect('equal')
plt.show()
```

## 5. Física Matemática del Arcoíris

Los arcoíris son fenómenos ópticos generados por la refracción, reflexión y dispersión cromática de la luz solar al atravesar gotas de lluvia esféricas en la atmósfera.

### 5.1 Planteamiento Geométrico
Consideremos un rayo de sol que incide en una gota de agua esférica de índice $n$ con un ángulo de incidencia $\theta_1$ respecto a la normal del punto de contacto:
1. El rayo se refracta hacia el interior con un ángulo $\theta_2$ dado por la ley de Snell ($\sin\theta_1 = n \sin\theta_2$).
2. Viaja dentro de la gota y choca contra la pared posterior, donde parte de la energía se refleja hacia adentro con ángulo $\theta_2$.
3. Emerge refractándose de nuevo al aire con ángulo $\theta_1$.

La desviación angular acumulada del rayo saliente con respecto a la dirección solar original se denota por el ángulo de desviación $2\phi$, donde la mitad del ángulo de visión $\phi$ está dado geométricamente por:
$$\phi = 2\theta_2 - \theta_1 \quad \text{(Ecuación 1)}$$
En la literatura, se suele expresar en términos del ángulo de reflexión interna $\beta$ (equivalente a $\theta_2$). Aplicando la ley de Snell en términos de $\phi$ y $\beta$:
$$\sin(2\beta - \phi) = n \sin\beta \implies \phi = 2\beta - \arcsin(n \sin\beta) \quad \text{(Ecuación 2)}$$

### 5.2 Determinación del Ángulo de Intensidad Máxima
El arcoíris se forma en el ángulo donde la luz reflejada alcanza su máxima concentración espacial. Esto corresponde matemáticamente al punto estacionario de la desviación angular (máximo local), donde la derivada de $\phi$ respecto al ángulo interno $\beta$ es nula:
$$\frac{d\phi}{d\beta} = 0 \implies 2 - \frac{d}{d\beta}\left[ \arcsin(n \sin\beta) \right] = 0$$
Sabiendo que la derivada del arcoseno es $\frac{d}{dx}[\arcsin(x)] = \frac{1}{\sqrt{1-x^2}}$, y aplicando la regla de la cadena:
$$2 - \frac{n \cos\beta}{\sqrt{1 - n^2 \sin^2\beta}} = 0 \implies \frac{n \cos\beta}{\sqrt{1 - n^2 \sin^2\beta}} = 2 \quad \text{(Ecuación 3)}$$

> [!WARNING]
> **Corrección de Errata en la Literatura del Curso:**
> En la guía de estudio del curso (páginas 12-13), el autor introduce una errata al sustituir la identidad de $\sin^2\beta$. Reemplaza erróneamente $\sin^2\beta = (1 - \cos\beta)^2$, lo que le conduce a la ecuación cuadrática incorrecta:
> $$5n^2 \cos^2\beta - 8n^2 \cos\beta + (4n^2 - 4) = 0 \quad \text{(Ecuación Errónea)}$$
> 
> **Deducción Correcta:**
> Aplicando la identidad pitagórica correcta, $\sin^2\beta = 1 - \cos^2\beta$, en la Ecuación 3 elevada al cuadrado:
> $$\frac{n^2 \cos^2\beta}{1 - n^2 \sin^2\beta} = 4 \implies n^2 \cos^2\beta = 4\left(1 - n^2 [1 - \cos^2\beta]\right)$$
> $$n^2 \cos^2\beta = 4 - 4n^2 + 4n^2 \cos^2\beta \implies 3n^2 \cos^2\beta = 4n^2 - 4$$
> De donde se obtiene la **ecuación simplificada correcta**:
> $$\cos\beta_{\text{max}} = \sqrt{\frac{4(n^2 - 1)}{3n^2}} = \frac{2\sqrt{n^2 - 1}}{\sqrt{3}n}$$
> Note que aunque el texto del curso comete la errata algebraica intermedia en la cuadrática, presenta esta misma fórmula final como resultado.

### 5.3 Cálculo Numérico para el Agua
Para el agua ($n \approx 1.333$):
$$\cos\beta_{\text{max}} = \frac{2\sqrt{1.333^2 - 1}}{\sqrt{3} \times 1.333} \approx 0.763 \implies \beta_{\text{max}} \approx 40.2^\circ$$
El ángulo de incidencia correspondiente en la entrada es:
$$\theta_{1,\text{max}} = 2\beta_{\text{max}} - \phi_{\text{max}} \implies \sin\theta_{1,\text{max}} = n \sin\beta_{\text{max}} \approx 1.333 \sin(40.2^\circ) \approx 0.860 \implies \theta_{1,\text{max}} \approx 59.4^\circ$$
Por lo tanto, la mitad del ángulo de visión $\phi_{\text{max}}$ es:
$$\phi_{\text{max}} = 2(40.2^\circ) - 59.4^\circ = 21.0^\circ$$
El ángulo completo de visión del arcoíris primario respecto al eje de la sombra del observador es de:
$$2\phi_{\text{max}} = 2(21.0^\circ) \approx 42.0^\circ$$
Debido a la dispersión, el índice del agua para la luz roja es $n \approx 1.331$ ($2\phi \approx 42.3^\circ$) y para la luz azul $n \approx 1.344$ ($2\phi \approx 40.4^\circ$). Esto genera la banda de colores con el rojo en el borde exterior y el azul en el interior.


```python
# 1. Graficar el ángulo de salida del arcoíris como función del parámetro de incidencia
n_water = 1.333
theta1_vals = np.linspace(0, np.radians(89.0), 300)
theta2_vals = np.arcsin(np.sin(theta1_vals) / n_water)
phi_vals = 4 * theta2_vals - 2 * theta1_vals  # Ángulo de visión final de la luz

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.plot(np.degrees(theta1_vals), np.degrees(phi_vals), 'r-', linewidth=2.5, label='Desviación angular')
ax1.axhline(42.0, color='black', linestyle='--', label='Ángulo Límite del Arcoíris ~42°')
ax1.set_title('Ángulo del Arcoíris en función de la Incidencia', fontsize=12, fontweight='bold')
ax1.set_xlabel('Ángulo de Incidencia $\\theta_1$ (grados)', fontsize=11)
ax1.set_ylabel('Ángulo de Emergencia $2\\phi$ (grados)', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.legend(frameon=True)

# 2. Trazado de rayos dentro de una gota esférica 2D
R_drop = 1.0
y_impacts = np.linspace(0.05, 0.95, 25)  # Parámetros de impacto b = R*sin(theta1)

# Dibujar la gota de agua circular
circle_t = np.linspace(0, 2*np.pi, 200)
ax2.plot(R_drop * np.cos(circle_t), R_drop * np.sin(circle_t), 'deepskyblue', linewidth=3, zorder=2)
ax2.fill(R_drop * np.cos(circle_t), R_drop * np.sin(circle_t), 'lightcyan', alpha=0.3)

for yi in y_impacts:
    # Punto de incidencia en la cara izquierda
    theta1 = np.arcsin(yi / R_drop)
    x_in = -R_drop * np.cos(theta1)
    y_in = yi
    
    # Ley de Snell
    theta2 = np.arcsin(np.sin(theta1) / n_water)
    
    # Punto de reflexión en la parte posterior
    alpha_refl = 2 * theta2 - theta1
    x_refl = R_drop * np.cos(alpha_refl)
    y_refl = R_drop * np.sin(alpha_refl)
    
    # Punto de salida en la cara inferior izquierda
    alpha_out = 4 * theta2 - theta1 - np.pi
    x_out = R_drop * np.cos(alpha_out)
    y_out = R_drop * np.sin(alpha_out)
    
    # Dibujar caminos de rayos
    ax2.plot([-2.0, x_in], [y_in, y_in], 'y-', alpha=0.4)  # Rayo solar entrante
    ax2.plot([x_in, x_refl], [y_in, y_refl], 'b-', alpha=0.6)  # Refracción
    ax2.plot([x_refl, x_out], [y_refl, y_out], 'b-', alpha=0.6)  # Reflexión interna
    
    # Rayo saliente emergente
    phi_val = 4 * theta2 - 2 * theta1
    x_exit_track = [x_out, x_out - 1.5 * np.cos(phi_val)]
    y_exit_track = [y_out, y_out - 1.5 * np.sin(phi_val)]
    ax2.plot(x_exit_track, y_exit_track, 'r-', alpha=0.5)

ax2.set_title('Concentración de Rayos en el Borde de Descartes (42°)', fontsize=12, fontweight='bold')
ax2.set_xlim(-1.8, 1.2)
ax2.set_ylim(-1.5, 1.2)
ax2.grid(True, linestyle=':', alpha=0.3)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## 6. Polarización y Birrefringencia

La polarización, como se ha visto, es la propiedad geométrica de las oscilaciones del campo eléctrico. En la atmósfera, la luz solar es polarizada parcialmente al dispersarse por colisiones con moléculas de aire y polvo (dispersión de Rayleigh), fenómeno que da origen al azul del cielo.

### 6.1 Birrefringencia (Doble Refracción)
La **birrefringencia** o doble refracción es una propiedad óptica presente en materiales anisótropos (como la calcita o plásticos bajo estrés mecánico), donde el índice de refracción del material depende de la polarización de la luz y su dirección de propagación. Un rayo de luz no polarizado incidente sobre un cristal birrefringente se divide en dos rayos que viajan a diferentes velocidades y se refractan con ángulos diferentes:

1. **Rayo Ordinario ($o$-ray):** Sigue la ley de Snell clásica de manera estricta. Su campo eléctrico vibra de forma perpendicular al eje óptico del cristal, experimentando un índice de refracción constante $n_o$.
2. **Rayo Extraordinario ($e$-ray):** No sigue la ley de Snell en su forma clásica (puede desviarse incluso con incidencia normal). Su campo eléctrico oscila en el plano que contiene al eje óptico, experimentando un índice de refracción $n_e$ variable con el ángulo.


```python
# Simulación de doble refracción en un cristal de Calcita (n_o = 1.658, n_e = 1.486)

n_o = 1.658
n_e = 1.486
theta_i_crystal = np.radians(20.0)

# Primera cara del cristal (Aire -> Cristal)
sin_theta_o = np.sin(theta_i_crystal) / n_o
theta_o = np.arcsin(sin_theta_o)

# El rayo extraordinario experimenta una refracción anisotrópica desviada adicionalmente
sin_theta_e = np.sin(theta_i_crystal) / n_e
theta_e = np.arcsin(sin_theta_e)

# Dibujar el cristal y las trayectorias de los dos rayos polarizados
plt.figure(figsize=(10, 6))
plt.axvline(0, color='black', linestyle='-', linewidth=2, label='Cara de entrada del Cristal')
plt.axvline(1.5, color='black', linestyle='-', linewidth=2, label='Cara de salida del Cristal')
plt.fill_between([0, 1.5], -1.0, 1.0, color='lightgray', alpha=0.3, label='Cristal de Calcita')

# Rayo incidente
x_inc_c = [-0.5, 0.0]
y_inc_c = [-0.5 * np.tan(theta_i_crystal), 0.0]
plt.plot(x_inc_c, y_inc_c, 'k-', linewidth=2.5, label='Rayo Incidente (No polarizado)')

# Rayo ordinario (Polarización perpendicular)
x_o = [0.0, 1.5]
y_o = [0.0, 1.5 * np.tan(theta_o)]
plt.plot(x_o, y_o, 'b-', linewidth=2, label=f'Rayo Ordinario o-ray (n_o={n_o})')

# Rayo extraordinario (Polarización paralela)
x_e = [0.0, 1.5]
y_e = [0.0, 1.5 * np.tan(theta_e)]
plt.plot(x_e, y_e, 'r-', linewidth=2, label=f'Rayo Extraordinario e-ray (n_e={n_e})')

plt.title('Doble Refracción (Birrefringencia) en un Cristal de Calcita', fontsize=13, fontweight='bold')
plt.xlabel('Dimensión longitudinal X (m)', fontsize=11)
plt.ylabel('Dimensión transversal Y (m)', fontsize=11)
plt.xlim(-0.8, 2.0)
plt.ylim(-0.8, 1.2)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper left')
plt.show()
```

## 7. Verificación de Resultados Numéricos

En esta sección de verificación validamos analíticamente la deducción del ángulo máximo de la gota del arcoíris primario mediante cálculos de optimización y aserciones.


```python
from scipy.optimize import minimize_scalar

# Definir la función del ángulo de desviación del arcoíris primario
def rainbow_angle_fun(theta1, n):
    # theta1 en radianes, n es el índice de refracción
    theta2 = np.arcsin(np.sin(theta1) / n)
    # El ángulo de visión es 2 * phi = 2 * (4 * theta2 - 2 * theta1)
    return 2 * (4 * theta2 - 2 * theta1)

# Encontramos el máximo del ángulo de visión
# Minimizamos el negativo de la función en la vecindad del máximo [50, 70] grados
n_val = 1.333
res_rb = minimize_scalar(lambda x: -rainbow_angle_fun(x, n_val), bounds=(np.radians(50.0), np.radians(70.0)), method='bounded')

theta1_max_num = res_rb.x
phi_max_num = -res_rb.fun

# Validación analítica con la ecuación de Descartes corregida:
cos_theta2_max = np.sqrt((n_val**2 - 1) / 3) / n_val if n_val > 1.0 else 1.0  # de la simplificación correcta de 3*n^2*cos^2 = 4*n^2 - 4
# Note que cos_theta2_max es igual a: 2 * sqrt(n^2 - 1) / (sqrt(3) * n) debido a la geometría
cos_theta2_calc = (2 * np.sqrt(n_val**2 - 1)) / (np.sqrt(3) * n_val)
theta2_max_theory = np.arccos(cos_theta2_calc)
theta1_max_theory = np.arcsin(n_val * np.sin(theta2_max_theory))
phi_max_theory = 2 * (4 * theta2_max_theory - 2 * theta1_max_theory)

print('--- VERIFICACIÓN DE LA FÍSICA DEL ARCOÍRIS ---')
print(f'Ángulo de incidencia de Descartes (Numérico): {np.degrees(theta1_max_num):.4f}°')
print(f'Ángulo de incidencia de Descartes (Teórico): {np.degrees(theta1_max_theory):.4f}°')
print(f'Ángulo de visión máximo del Arcoíris (Numérico): {np.degrees(phi_max_num):.4f}°')
print(f'Ángulo de visión máximo del Arcoíris (Teórico): {np.degrees(phi_max_theory):.4f}°')

assert np.isclose(theta1_max_num, theta1_max_theory, atol=1e-5)
assert np.isclose(phi_max_num, phi_max_theory, atol=1e-5)
print('Verificaciones del Arcoíris: ¡EXITOSA! (Los resultados analíticos y numéricos coinciden perfectamente)')
```

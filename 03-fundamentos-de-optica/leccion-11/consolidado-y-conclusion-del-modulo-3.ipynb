## Fase 1: Emisión y Propagación de Ondas Electromagnéticas (MAS, Polarización y Poynting)

En el origen (la estrella), una transición atómica genera una corriente de carga oscilante. Desde el punto de vista clásico, esto corresponde a un dipolo eléctrico que oscila en Movimiento Armónico Simple (MAS) (Lección 1). Esta oscilación perturba el campo electromagnético local y propaga una onda electromagnética en el vacío (Lección 5).

A partir de las **Ecuaciones de Maxwell en el vacío** (con densidad de carga $\rho=0$ y densidad de corriente $\vec{J}=0$):
$$\nabla \cdot \vec{E} = 0, \quad \nabla \cdot \vec{B} = 0$$
$$\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}, \quad \nabla \times \vec{B} = \mu_0 \epsilon_0 \frac{\partial \vec{E}}{\partial t}$$

Aplicando el operador rotacional a las ecuaciones de rotación y usando la identidad vectorial $\nabla \times (\nabla \times \vec{A}) = \nabla(\nabla \cdot \vec{A}) - \nabla^2 \vec{A}$, derivamos la **ecuación de onda electromagnética** (Lección 2, Lección 5):
$$\nabla^2 \vec{E} - \frac{1}{c^2} \frac{\partial^2 \vec{E}}{\partial t^2} = 0$$
Donde la velocidad de la luz en el vacío es $c = \frac{1}{\sqrt{\mu_0 \epsilon_0}} \approx 3 \times 10^8 \text{ m/s}$.

### 1.1 Solución de Onda Plana y Estados de Polarización
Para una onda armónica plana que viaja en la dirección $z$, el campo eléctrico se puede expresar como:
$$\vec{E}(z, t) = E_{0x} \cos(kz - \omega t + \phi_x) \hat{i} + E_{0y} \cos(kz - \omega t + \phi_y) \hat{j}$$
El campo magnético correspondiente es perpendicular y está en fase, dado por:
$$\vec{B}(z, t) = \frac{1}{c} (\hat{k} \times \vec{E}(z,t))$$
La diferencia de fase $\Delta \phi = \phi_y - \phi_x$ determina el estado de polarización (Lección 2, Lección 5, Lección 6):
- **Lineal:** $\Delta \phi = 0 \text{ ó } \pi$.
- **Circular:** $\Delta \phi = \pm \pi/2$ y $E_{0x} = E_{0y}$.
- **Elíptica:** Cualquier otro valor general.

### 1.2 Flujo de Energía: Vector de Poynting
La energía transportada por unidad de área y tiempo se describe mediante el **Vector de Poynting**:
$$\vec{S} = \frac{1}{\mu_0} \vec{E} \times \vec{B}$$
Para una onda plana, la magnitud promedio en el tiempo (intensidad) es $I = \langle S \rangle = \frac{1}{2} c \epsilon_0 E_0^2$.


````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Configurar estilo gráfico
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# =======================================================================
# 1. VALIDACIÓN SIMBÓLICA CON SYMPY: Ecuación de Faraday de Maxwell
# =======================================================================
z_sym, t_sym, E0, B0_sym, k_sym, w_sym, c_sym = sp.symbols('z t E0 B0 k w c')
# Definimos campos para onda polarizada linealmente en el eje x
Ex = E0 * sp.cos(k_sym * z_sym - w_sym * t_sym)
By = B0_sym * sp.cos(k_sym * z_sym - w_sym * t_sym)

# E_vec = (Ex, 0, 0) y B_vec = (0, By, 0)
# Faraday: rot(E) = -dB/dt
# rot(E)_y = dEx/dz - dEz/dx = dEx/dz
# -dB/dt_y = -dBy/dt
rot_E_y = sp.diff(Ex, z_sym)
neg_dB_dt_y = -sp.diff(By, t_sym)

print("--- VALIDACIÓN SIMBÓLICA DE MAXWELL (Fase 1) ---")
print(f"Componente y de rot(E): {rot_E_y}")
print(f"Componente y de -dB/dt: {neg_dB_dt_y}")
# Relación para satisfacer Faraday:
# -k * E0 * sin(...) == -w * B0 * sin(...) => k * E0 = w * B0 => E0/B0 = w/k
relacion = sp.solve(rot_E_y - neg_dB_dt_y, B0_sym)
print(f"Campo magnético necesario B0 en términos de E0, w, k: {relacion[0]}")
print(f"Dado que w/k = c, se verifica que E0/B0 = c.\n")

# =======================================================================
# 2. SIMULACIÓN DE POLARIZACIONES DE LA LUZ
# =======================================================================
t_vals = np.linspace(0, 2*np.pi, 200)
E_0x = 1.0
E_0y = 1.0

# Diferencias de fase para distintas polarizaciones
fases = {
    'Lineal (45°)': 0,
    'Circular Derecha': -np.pi/2,
    'Elíptica': -np.pi/4
}

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, (name, delta) in zip(axes, fases.items()):
    Ex_vals = E_0x * np.cos(-t_vals)  # z = 0
    Ey_vals = E_0y * np.cos(-t_vals + delta)
    
    ax.plot(Ex_vals, Ey_vals, lw=2.5, color='crimson' if 'Lineal' in name else 'forestgreen' if 'Circular' in name else 'darkblue')
    ax.quiver(0, 0, Ex_vals[0], Ey_vals[0], scale=2.5, color='black', alpha=0.3)
    ax.quiver(0, 0, Ex_vals[50], Ey_vals[50], scale=2.5, color='red', alpha=0.8, label='Vector Campo E')
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.set_aspect('equal')
    ax.set_title(name, fontsize=12, fontweight='bold')
    ax.set_xlabel('E_x')
    ax.set_ylabel('E_y')
    ax.axhline(0, color='gray', lw=0.5, ls='--')
    ax.axvline(0, color='gray', lw=0.5, ls='--')
    ax.legend(loc='upper right')

plt.suptitle(r'Estados de Polarización del Campo Eléctrico $\vec{E}$ en el plano transversal ($z=0$)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## Fase 2: Propagación e Interacción Atmosférica (Dispersión y Acústica)

El haz de luz viaja a través del espacio y entra en la atmósfera de la Tierra. El aire no es un vacío: es un medio dieléctrico dispersivo con un índice de refracción $n(\lambda)$ que depende de la longitud de onda de la luz. Este comportamiento se modela mediante la **Ecuación dispersiva de Cauchy** (Lección 6):
$$n(\lambda) = A + \frac{B}{\lambda^2}$$
Donde para el aire, los coeficientes típicos son $A \approx 1.00027$ y $B \approx 2.5 \times 10^{-18} \text{ m}^2$.

### 2.1 Refracción Acusto-Óptica (Efecto del Sonido en la Luz)
Las ondas de sonido son variaciones longitudinales de presión y densidad en un gas ideal (Lección 4). El sonido viaja con una velocidad adiabática dada por:
$$v_s = \sqrt{\frac{\gamma R T}{M}}$$
Donde $\gamma$ es el coeficiente adiabático, $R$ es la constante de los gases, $T$ es la temperatura y $M$ la masa molar del gas.

Estas oscilaciones de densidad producen variaciones periódicas en el índice de refracción del aire (Acousto-optics), comportándose como una rejilla de índice dinámica:
$$n(x, t) = n_0 + \Delta n \cos(k_s x - \omega_s t)$$
Donde $k_s$ y $\omega_s$ son el número de onda y la frecuencia angular de la onda sonora.

Un rayo de luz que entra perpendicularmente a la propagación del sonido sufrirá refracción continua por la variación del índice de refracción. De acuerdo con la ecuación diferencial de trazado de rayos en medios inhomogéneos (derivada de la **Ley de Snell** infinitesimal, Lección 6, Lección 7):
$$\frac{d^2 x}{dy^2} \approx \frac{1}{n_0} \frac{\partial n}{\partial x}$$
Esto causa que el rayo de luz se curve de forma senoidal a medida que se propaga longitudinalmente en el eje $y$.


````python
# =======================================================================
# 1. MODELADO DE LA DISPERSIÓN DE CAUCHY
# =======================================================================
A = 1.000272  # Parámetro A para el aire
B = 2.45e-18  # Parámetro B para el aire (m^2)
lambdas = np.linspace(380e-9, 750e-9, 500)  # Espectro visible
n_cauchy = A + B / lambdas**2

plt.figure(figsize=(8, 4.5))
plt.plot(lambdas * 1e9, n_cauchy, lw=2.5, color='darkorange', label=r'Ecuación de Cauchy (Aire)')
plt.title('Dispersión Cromática en la Atmósfera', fontsize=12, fontweight='bold')
plt.xlabel(r'Longitud de Onda $\lambda$ (nm)')
plt.ylabel(r'Índice de Refracción $n(\lambda)$')
plt.legend()
plt.show()

# =======================================================================
# 2. TRAZADO DE RAYOS EN UN GRADIENTE DE ÍNDICE INDUCIDO POR SONIDO
# =======================================================================
# Supongamos una onda de sonido estacionaria en el eje x con longitud de onda lambda_s = 2 mm
lambda_s = 2.0e-3 # 2 mm
ks = 2 * np.pi / lambda_s
Dn = 5.0e-5  # Modulación del índice de refracción
n0 = 1.00027

# El rayo entra en y=0, x=x0 y se propaga en y. Pasos de integración dy
y_max = 50e-3  # 50 mm de propagación
N_pasos = 1000
y_vals = np.linspace(0, y_max, N_pasos)
dy = y_max / N_pasos

# Simulamos para diferentes posiciones de entrada x0
x0_list = np.linspace(-lambda_s/4, lambda_s/4, 5)

plt.figure(figsize=(10, 6))
for x0 in x0_list:
    x_ray = np.zeros(N_pasos)
    vx_ray = np.zeros(N_pasos)  # dx/dy
    x_ray[0] = x0
    vx_ray[0] = 0.0  # Incidencia perpendicular
    
    for i in range(1, N_pasos):
        # Ecuación del gradiente: dn/dx = -Dn * ks * sin(ks * x)
        dn_dx = -Dn * ks * np.sin(ks * x_ray[i-1])
        # Aceleración transversal: d^2x/dy^2 = (1/n0) * dn/dx
        ax = dn_dx / n0
        
        # Integrador numérico de Euler
        vx_ray[i] = vx_ray[i-1] + ax * dy
        x_ray[i] = x_ray[i-1] + vx_ray[i] * dy
        
    plt.plot(y_vals * 1e3, x_ray * 1e3, lw=2, label=rf'$x_0 = {x0*1e3:.2f}$ mm')

plt.title('Deflexión y Trazado de Rayos de Luz por Ondas Acústicas (Interacción Sonido-Luz)', fontsize=12, fontweight='bold')
plt.xlabel('Distancia de Propagación $y$ (mm)')
plt.ylabel('Desviación Transversal $x$ (mm)')
plt.legend(loc='upper right')
plt.show()
````

## Fase 3: Enfoque Óptico y Corrección de Aberraciones (Matrices ABCD y Doblete Acromático)

Para recoger la débil luz estelar, el telescopio utiliza un objetivo compuesto por lentes. Modelamos el trazado de rayos mediante el **formalismo de matrices de transferencia óptica ABCD** de la Óptica Geométrica (Lección 7, Lección 8).

Un rayo de luz en un plano transversal a una distancia del eje óptico $r$ y con un ángulo de inclinación $\theta$ se describe como el vector $\begin{pmatrix} r \\ \theta \end{pmatrix}$. El paso del rayo a través de elementos ópticos se realiza multiplicando sus matrices de transferencia correspondiente:
- **Espacio libre de distancia $d$:**
  $$\mathbf{M}_{libre} = \begin{pmatrix} 1 & d \\ 0 & 1 \end{pmatrix}$$
- **Lente delgada con distancia focal $f$:**
  $$\mathbf{M}_{lente} = \begin{pmatrix} 1 & 0 \\ -1/f & 1 \end{pmatrix}$$
- **Refracción en interfaz esférica con radio de curvatura $R$ de un medio $n_1$ a $n_2$:**
  $$\mathbf{M}_{ref} = \begin{pmatrix} 1 & 0 \\ \frac{n_1 - n_2}{R n_2} & \frac{n_1}{n_2} \end{pmatrix}$$

### 3.1 Aberración Cromática y el Doblete Acromático
Debido a la dispersión de Cauchy, el índice de refracción del vidrio de la lente depende de la longitud de onda ($\lambda$). Esto causa que la distancia focal $f$ sea distinta para cada color, provocando un desenfoque cromático (Aberración Cromática, Lección 8):
$$\frac{1}{f(\lambda)} = (n(\lambda) - 1) \left( \frac{1}{R_1} - \frac{1}{R_2} \right)$$

Para corregir esta aberración en el espectro visible, diseñamos un **doblete acromático** en contacto, compuesto por:
1. Una lente convergente de vidrio Crown (baja dispersión, número de Abbe $V_1$ alto).
2. Una lente divergente de vidrio Flint (alta dispersión, número de Abbe $V_2$ bajo).

El número de Abbe $V$ se define como:
$$V = \frac{n_{C} - 1}{n_F - n_C}$$
Donde para nuestro diseño elegiremos las longitudes de onda correspondientes a la línea C del Hidrógeno (rojo, 656.3 nm) y la línea F (azul, 486.1 nm).

Las potencias ópticas individuales ($P_1 = 1/f_1$ y $P_2 = 1/f_2$) necesarias para obtener una potencia total deseada $P = 1/f_{tot}$ libre de aberración cromática en C y F satisfacen las ecuaciones:
$$P_1 + P_2 = P$$
$$\frac{P_1}{V_1} + \frac{P_2}{V_2} = 0$$
Resolviendo este sistema lineal de ecuaciones, calculamos las distancias focales exactas de cada elemento del doblete.


````python
# =======================================================================
# 1. DISEÑO DEL DOBLETE ACROMÁTICO
# =======================================================================
# Índices de refracción para vidrio Crown (Lente 1) y vidrio Flint (Lente 2)
# En longitudes de onda: Rojo (656 nm), Verde (587 nm), Azul (486 nm)
n_crown = {'R': 1.514, 'V': 1.517, 'A': 1.523}
n_flint = {'R': 1.615, 'V': 1.621, 'A': 1.635}

# Cálculo de números de Abbe
V_crown = (n_crown['V'] - 1) / (n_crown['A'] - n_crown['R'])
V_flint = (n_flint['V'] - 1) / (n_flint['A'] - n_flint['R'])

print("--- NÚMEROS DE ABBE PARA DISEÑO ÓPTICO ---")
print(f"Número de Abbe Vidrio Crown (bajo dispersivo): {V_crown:.2f}")
print(f"Número de Abbe Vidrio Flint (alto dispersivo): {V_flint:.2f}\n")

# Distancia focal objetivo del doblete acromático: f_tot = 100 mm
f_tot = 0.100  # 100 mm
P_tot = 1.0 / f_tot

# Resolver sistema lineal de ecuaciones para potencias P1 y P2:
# P1 + P2 = P_tot
# P1/V_1 + P2/V_2 = 0 => P2 = -P1 * (V_2 / V_1)
# P1 * (1 - V_2/V_1) = P_tot => P1 = P_tot / (1 - V_2/V_1)
P1 = P_tot / (1.0 - V_flint / V_crown)
P2 = P_tot - P1
f1 = 1.0 / P1
f2 = 1.0 / P2

print(f"Distancias focales necesarias:")
print(f"Lente 1 (Crown): {f1 * 1000:.2f} mm (Convergente)")
print(f"Lente 2 (Flint): {f2 * 1000:.2f} mm (Divergente)\n")

# =======================================================================
# 2. TRAZADO DE RAYOS MULTI-COLOR (ABCD) PARA LENTE SIMPLE VS DOBLETE
# =======================================================================
def trazar_sistema(f_vals, d_detector, r_entrada=0.010, color='r'):
    """
    Traza un rayo de altura inicial r_entrada a través de lentes en contacto
    y se propaga hasta el detector.
    """
    # Vector de entrada del rayo
    v_ray = np.array([r_entrada, 0.0])
    
    # Matriz del sistema de lentes (potencia combinada P = sum(1/f))
    P_combo = sum([1.0/f for f in f_vals])
    M_lente = np.array([[1.0, 0.0], [-P_combo, 1.0]])
    
    # Pasar por la lente
    v_lente = M_lente @ v_ray
    
    # Trazar propagación en espacio libre tras la lente en función de la distancia z
    z_steps = np.linspace(0, 0.15, 200) # De 0 a 150 mm
    ray_heights = []
    for z in z_steps:
        M_libre = np.array([[1.0, z], [0.0, 1.0]])
        ray_heights.append((M_libre @ v_lente)[0])
        
    return z_steps, np.array(ray_heights)

# Comparación de trayectoria cromática
colores_luz = {'R': 'red', 'V': 'green', 'A': 'blue'}
alturas_iniciales = [0.005, 0.010]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 8.5))

# 1. LENTE SIMPLE (Solo Crown, focal de 100 mm nominal en Verde)
for c_code, c_plot in colores_luz.items():
    # Índice de refracción dependiente del color en Crown
    n_c = n_crown[c_code]
    # f_color = f_nominal * (n_nominal - 1) / (n_color - 1)
    f_color = 0.100 * (n_crown['V'] - 1) / (n_c - 1)
    for r0 in alturas_iniciales:
        z_s, heights = trazar_sistema([f_color], 0.1, r_entrada=r0)
        ax1.plot(z_s * 1000, heights * 1000, color=c_plot, alpha=0.7)
ax1.axvline(100, color='black', ls='--', alpha=0.5, label='Foco Nominal (100 mm)')
ax1.set_title('Aberración Cromática en una Lente Simple (Crown)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Altura del Rayo $r$ (mm)')
ax1.legend()

# 2. DOBLETE ACROMÁTICO CROWN-FLINT
for c_code, c_plot in colores_luz.items():
    # f_color = f_nominal * (n_nominal - 1) / (n_color - 1) para cada lente
    f1_c = f1 * (n_crown['V'] - 1) / (n_crown[c_code] - 1)
    f2_c = f2 * (n_flint['V'] - 1) / (n_flint[c_code] - 1)
    for r0 in alturas_iniciales:
        z_s, heights = trazar_sistema([f1_c, f2_c], 0.1, r_entrada=r0)
        ax2.plot(z_s * 1000, heights * 1000, color=c_plot, alpha=0.7)
ax2.axvline(100, color='black', ls='--', alpha=0.5, label='Foco Corregido (100 mm)')
ax2.set_title('Corrección Cromática Mediante el Doblete Acromático Diseñado', fontsize=12, fontweight='bold')
ax2.set_xlabel(r'Distancia del Eje Óptico $z$ (mm)')
ax2.set_ylabel('Altura del Rayo $r$ (mm)')
ax2.legend()

plt.tight_layout()
plt.show()
````

## Fase 4: Modulación Interferométrica (Película Antirreflejante e Interferómetro de Michelson)

Una vez enfocado el haz de luz, aplicamos dos técnicas de interferometría física para optimizar su detección y medir sus propiedades ondulatorias.

### 4.1 Película Delgada Antirreflejante sobre la Lente
Para evitar pérdidas de luz por reflexión en la superficie de vidrio de la primera lente ($n_g = 1.52$), se deposita una película delgada de Fluoruro de Magnesio ($MgF_2$, $n_f = 1.38$) (Lección 9).

Dado que $n_{aire} < n_f < n_g$, la onda reflejada en la interfaz aire-película experimenta un salto de fase de $\pi$ radianes. Del mismo modo, la onda que se refleja en la interfaz película-vidrio también experimenta un salto de fase de $\pi$.

Para que la reflexión sea nula por interferencia destructiva en incidencia normal, la diferencia de camino óptico debe corresponder a media longitud de onda:
$$2 n_f t = \frac{\lambda_0}{2} \implies t_{opt} = \frac{\lambda_0}{4 n_f}$$
Donde $\lambda_0$ es la longitud de onda de diseño óptimo (típicamente 550 nm en el verde, máximo de sensibilidad del ojo humano).

### 4.2 El Interferómetro de Michelson
El haz atraviesa un divisor de haz y se separa en dos brazos con longitudes $L_1$ y $L_2$ (Lección 9). Al reflejarse de nuevo y recombinarse, la diferencia de camino óptico es $\Delta d = 2(L_1 - L_2)$. Si el haz no está perfectamente alinea o incide con cierto ángulo $\theta$ respecto a la normal del espejo, la intensidad de interferencia proyectada en una pantalla forma un patrón de **anillos concéntricos** (franjas circulares):
$$I(\theta) = I_0 \cos^2 \left( \frac{\pi \Delta d \cos\theta}{\lambda} \right)$$


````python
# =======================================================================
# 1. MODELADO DE REFLECTIVIDAD DE PELÍCULA DELGADA ANTIRREFLEJANTE
# =======================================================================
lambda_diseno = 550e-9  # 550 nm (Verde)
n_f = 1.38  # MgF2
n_g = 1.52  # Vidrio Crown

# Grosor de la película antirreflejante de cuarto de onda (t = lambda / 4n_f)
t_film = lambda_diseno / (4 * n_f)
print("--- DISEÑO DE PELÍCULA DELGADA ANTIRREFLEJANTE (Fase 4) ---")
print(f"Espesor óptimo de la película de MgF2: {t_film * 1e9:.2f} nm\n")

def calcular_reflectancia_thin_film(lam_vals, t, nf, ng):
    # Coeficientes de Fresnel para interfaces (aire-película y película-vidrio)
    n0 = 1.0
    r1 = (n0 - nf) / (n0 + nf)
    r2 = (nf - ng) / (nf + ng)
    # Diferencia de fase al atravesar la película (espesor t)
    delta = (2 * np.pi * nf * t) / lam_vals
    # Amplitud reflejada total compleja considerando múltiples reflexiones
    r_total = (r1 + r2 * np.exp(-2j * delta)) / (1 + r1 * r2 * np.exp(-2j * delta))
    return np.abs(r_total)**2

lam_plot = np.linspace(400e-9, 700e-9, 500)
R_con_pelicula = calcular_reflectancia_thin_film(lam_plot, t_film, n_f, n_g)
R_sin_pelicula = ((1.0 - n_g) / (1.0 + n_g))**2 * np.ones_like(lam_plot)

plt.figure(figsize=(8, 4.5))
plt.plot(lam_plot * 1e9, R_sin_pelicula * 100, 'k--', label='Vidrio sin recubrimiento')
plt.plot(lam_plot * 1e9, R_con_pelicula * 100, 'b-', lw=2.5, label=r'Con película de $MgF_2$ (550 nm)')
plt.title('Reflectancia de Superficie con Película Antirreflejante', fontsize=12, fontweight='bold')
plt.xlabel(r'Longitud de Onda $\lambda$ (nm)')
plt.ylabel('Reflectancia (%)')
plt.ylim(0, 5.0)
plt.legend()
plt.show()

# =======================================================================
# 2. SIMULACIÓN 2D DEL INTERFERÓMETRO DE MICHELSON (Anillos concéntricos)
# =======================================================================
lam_laser = 633e-9  # Láser rojo Helio-Neon
delta_d = 0.05e-3   # Diferencia de camino óptico L1 - L2 = 50 micrometros

# Generar grid de coordenadas angulares en la pantalla del detector
x = np.linspace(-0.05, 0.05, 400)
y = np.linspace(-0.05, 0.05, 400)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
# Relación angular theta (ángulo de incidencia en el espejo)
theta = np.arctan(R / 0.5)  # Pantalla colocada a L = 0.5 m del interferómetro

# Intensidad resultante
I_michelson = np.cos(np.pi * delta_d * np.cos(theta) / lam_laser)**2

plt.figure(figsize=(6, 6))
plt.imshow(I_michelson, cmap='gray', extent=[-5, 5, -5, 5])
plt.title(rf'Patrón de Anillos del Interferómetro de Michelson\n$\Delta d = {delta_d*1e6:.1f}$ $\mu m$', fontsize=12, fontweight='bold')
plt.xlabel('Posición Transversal X (cm)')
plt.ylabel('Posición Transversal Y (cm)')
plt.grid(False)
plt.show()
````

## Fase 5: Espectroscopía por Red de Difracción y Límite de Resolución (Rayleigh)

Finalmente, para descifrar la física química de la estrella, analizamos su espectro mediante un espectrómetro de red de difracción. La luz pasa por una rendija y luego incide en una **red de difracción de transmisión** de $N$ ranuras separadas por una distancia $d$ (Lección 10).

El perfil de intensidad difractada en la pantalla según la difracción de Fraunhofer y el principio de superposición combina la difracción por ranura individual de ancho $a$ y la interferencia de las $N$ ranuras coherentes (Lección 9, Lección 10):
$$I(\theta) = I_0 \left[ \frac{\sin(\beta)}{\beta} \right]^2 \left[ \frac{\sin(N \gamma)}{N \sin(\gamma)} \right]^2$$
Donde:
$$\beta = \frac{\pi a \sin\theta}{\lambda}, \quad \gamma = \frac{\pi d \sin\theta}{\lambda}$$

### 5.1 El Criterio de Rayleigh para Resolución Espectral
Para separar dos longitudes de onda muy cercanas (como el **doblete de emisión del Sodio** en la estrella: $\lambda_1 = 589.0 \text{ nm}$ y $\lambda_2 = 589.6 \text{ nm}$, $\Delta \lambda = 0.6 \text{ nm}$), el sistema de difracción debe ser capaz de resolverlas.

Según el **Criterio de Rayleigh**:
* *Dos líneas espectrales son justamente resolvedores si el máximo principal de orden $m$ de una longitud de onda coincide exactamente con el primer mínimo adyacente del mismo orden de la otra longitud de onda.*

Matemáticamente, esto se traduce en el poder de resolución límite para una red de difracción en el orden de difracción $m$:
$$R_P = \frac{\lambda}{\Delta \lambda} = m N$$
Donde $N$ es el número de líneas de la red iluminadas por el haz.

Para el doblete de Sodio en primer orden ($m=1$), el número de ranuras mínimo para resolverlo es:
$$N_{min} = \frac{\lambda}{\Delta \lambda} = \frac{589.3 \text{ nm}}{0.6 \text{ nm}} \approx 982 \text{ ranuras}$$
Si iluminamos menos de 982 ranuras, las dos líneas espectrales se fusionarán en un solo pico indistinguible. Si iluminamos más (por ejemplo, $N=2000$), las dos líneas se verán perfectamente resueltas como picos nítidos independientes.


````python
# =======================================================================
# SIMULACIÓN DEL PODER DE RESOLUCIÓN ESPECTRAL DEL DOBLETE DE SODIO
# =======================================================================
# Constantes de diseño física
a_slit = 2.5e-6   # Ancho de ranura = 2.5 micrometros
d_grating = 8.0e-6 # Espaciado de red = 8.0 micrometros (125 lineas/mm)

# Longitudes de onda del Doblete de Sodio D1 y D2
lam1 = 589.0e-9 # 589.0 nm
lam2 = 589.6e-9 # 589.6 nm

# Coordenadas angulares alrededor del máximo de difracción de primer orden (m=1)
# d*sin(theta) = m*lambda => sin(theta) = 589nm / 8um = 0.0736 => theta_1 = 4.22 grados
theta_c = np.arcsin(1.0 * lam1 / d_grating)
angulos = np.linspace(theta_c - 0.005, theta_c + 0.005, 1000)

def calcular_intensidad_red(theta_arr, lam, N, a, d):
    beta = (np.pi * a * np.sin(theta_arr)) / lam
    gamma = (np.pi * d * np.sin(theta_arr)) / lam
    
    # Evitar indeterminacion en beta=0
    factor_difrac = np.where(beta == 0, 1.0, (np.sin(beta)/beta)**2)
    factor_interf = np.where(np.sin(gamma) == 0, 1.0, (np.sin(N*gamma) / (N*np.sin(gamma)))**2)
    
    return factor_difrac * factor_interf

# Estudiamos dos casos de resolución según el número de ranuras N
# Caso A: N = 200 (No resuelto, menor que 982)
# Caso B: N = 1200 (Resuelto, mayor que 982)
N_casos = [200, 1200]
colors = ['crimson', 'forestgreen']

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for ax, N, col in zip(axes, N_casos, colors):
    # Calcular perfiles de intensidad individuales
    I_lam1 = calcular_intensidad_red(angulos, lam1, N, a_slit, d_grating)
    I_lam2 = calcular_intensidad_red(angulos, lam2, N, a_slit, d_grating)
    I_total = I_lam1 + I_lam2
    
    # Graficar contribuciones y total
    ax.plot(np.degrees(angulos), I_lam1, 'y--', alpha=0.6, label=r'Línea $D_1$ (589.0 nm)')
    ax.plot(np.degrees(angulos), I_lam2, 'orange', ls='--', alpha=0.6, label=r'Línea $D_2$ (589.6 nm)')
    ax.plot(np.degrees(angulos), I_total, color=col, lw=2.5, label='Intensidad Combinada Total')
    
    ax.set_title(f'Red con $N = {N}$ ranuras (m=1)\n¿Resuelto? {N >= 982}', fontsize=12, fontweight='bold')
    ax.set_xlabel(r'Ángulo de Difracción $	heta$ (grados)')
    ax.set_ylabel('Intensidad Normalizada (u.a.)')
    ax.legend(loc='upper right')
    ax.set_xlim(np.degrees(theta_c - 0.003), np.degrees(theta_c + 0.003))

plt.suptitle('Criterio de Rayleigh y Resolución del Doblete del Sodio Estelar', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

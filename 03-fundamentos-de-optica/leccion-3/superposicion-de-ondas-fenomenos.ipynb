## 1. Superposición de Ondas y el Experimento de Young

### 1.1 La Disputa Histórica: Newton vs. Huygens
Durante el siglo XVII y XVIII, la naturaleza física de la luz fue objeto de debate entre dos escuelas de pensamiento:
1. **Teoría Corpuscular (Isaac Newton):** Defendía que la luz consistía en una corriente de partículas rápidas (corpúsculos) que se desplazaban en línea recta.
2. **Teoría Ondulatoria (Christian Huygens):** Postulaba que la luz era una onda mecánica propagándose en un medio hipotético incorruptible que llenaba el vacío, denominado éter.

La teoría corpuscular dominó la física hasta 1801, año en el que el físico inglés **Thomas Young** realizó el histórico **experimento de la doble rendija**. Al iluminar una placa opaca con dos aperturas estrechas y paralelas separadas por una distancia $d$, observó sobre una pantalla colocada a una distancia $D$ un patrón alternante de franjas brillantes y oscuras. Este fenómeno de interferencia solo podía explicarse si la luz se comportaba como una onda.

### 1.2 Principio de Superposición Lineal
Cuando dos o más ondas coinciden en el mismo punto del espacio, la elongación resultante $y(x, t)$ es igual a la suma algebraica de las elongaciones individuales en ese instante:
$$y(x, t) = y_1(x, t) + y_2(x, t)$$

Una vez que las ondas se cruzan y continúan su viaje, recuperan exactamente sus formas e intensidades originales, demostrando que la interacción es puramente local y lineal.

### 1.3 Interferencia de Ondas Coherentes
Dos ondas son **coherentes** si poseen la misma amplitud $A$, longitud de onda $\lambda$, frecuencia $\omega$ y una diferencia de fase constante. Consideramos dos ondas que interfieren en un punto $P$:
$$F_1 = A \sin(\omega t - k x_1)$$
$$F_2 = A \sin(\omega t - k x_2)$$
Donde $x_1$ y $x_2$ representan las distancias desde los focos emisores hasta el punto de interferencia.

Aplicando la identidad trigonométrica de suma de senos:
$$\sin(X) + \sin(Y) = 2 \cos\left(\frac{X - Y}{2}\right) \sin\left(\frac{X + Y}{2}\right)$$

La onda superpuesta $F_T = F_1 + F_2$ es:
$$F_T = 2A \cos\left(\frac{k(x_2 - x_1)}{2}\right) \sin\left(\omega t - k \frac{x_1 + x_2}{2}\right)$$

La amplitud total de la interferencia $A_T$ en la posición del receptor es:
$$A_T = 2A \cos\left(\frac{k \Delta x}{2}\right)$$
Donde $\Delta x = x_2 - x_1$ es la diferencia de camino óptico.

> [!WARNING]
> **Corrección de Errata de la Literatura:**
> Algunos textos académicos aproximan o transcriben incorrectamente la amplitud resultante como $A_T = 2A \cos\left(\frac{2\pi}{\lambda} \Delta x\right)$. 
> La expresión correcta se obtiene reemplazando $k = 2\pi / \lambda$:
> $$\frac{k \Delta x}{2} = \frac{2\pi}{2\lambda} \Delta x = \frac{\pi \Delta x}{\lambda} \implies A_T = 2A \cos\left(\frac{\pi \Delta x}{\lambda}\right) = 2A \cos\left(\frac{\Delta \phi}{2}\right)$$
> Donde $\Delta \phi = k \Delta x$ es el desfase total.

* **Interferencia Constructiva (Máximos / Antinodos):** Ocurre cuando el coseno es $\pm 1$.
  $$\frac{\pi \Delta x}{\lambda} = n \pi \implies \Delta x = n \lambda, \quad n \in \{0, 1, 2, \dots\} \implies A_T = \pm 2A$$
* **Interferencia Destructiva (Mínimos / Nodos):** Ocurre cuando el coseno es $0$.
  $$\frac{\pi \Delta x}{\lambda} = (2n + 1)\frac{\pi}{2} \implies \Delta x = (2n + 1)\frac{\lambda}{2}, \quad n \in \{0, 1, 2, \dots\} \implies A_T = 0$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros físicos del experimento de Young
lam = 632.8e-9    # Longitud de onda del láser He-Ne (m)
d = 0.25e-3       # Distancia entre rendijas (m)
a = 0.04e-3       # Ancho de las rendijas (m)
D = 1.2           # Distancia a la pantalla (m)

# Posición en la pantalla (eje y vertical)
y_screen = np.linspace(-0.015, 0.015, 1000)

# 1. Patrón de interferencia pura (rendijas infinitesimales)
phi = (2 * np.pi / lam) * (d * y_screen / D)
I_interference = (np.cos(phi / 2))**2

# 2. Modulación por difracción de ranura simple (efecto sinc)
beta = (np.pi * a * y_screen) / (lam * D)
I_diffraction = (np.sinc(beta / np.pi))**2  # np.sinc(x) es sin(pi*x)/(pi*x)

# Intensidad total combinada
I_total = I_interference * I_diffraction

# Graficación
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [3, 1]})

# Curva de intensidad
ax1.plot(y_screen * 1000, I_total, 'r-', linewidth=2, label='Intensidad Combinada (Interferencia + Difracción)')
ax1.plot(y_screen * 1000, I_diffraction, 'k--', alpha=0.6, label='Envolvente de Difracción')
ax1.set_title('Perfil de Intensidad en la Pantalla (Experimento de Young)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Intensidad Normalizada I/I_0', fontsize=11)
ax1.grid(True, linestyle=':', alpha=0.6)
ax1.legend(frameon=True)

# Visualización simulada de las franjas (2D)
fringe_strip = np.tile(I_total, (100, 1))
ax2.imshow(fringe_strip, cmap='hot', aspect='auto', extent=[y_screen[0]*1000, y_screen[-1]*1000, 0, 1])
ax2.set_xlabel('Posición en la pantalla y (mm)', fontsize=11)
ax2.get_yaxis().set_visible(False)

plt.tight_layout()
plt.show()
```

## 2. Interferencia de Ondas No Coherentes: Batidos (Beats)

El fenómeno de los **batidos** o pulsaciones se produce cuando interfieren dos ondas de igual amplitud pero con frecuencias $\omega_1$ y $\omega_2$ ligeramente diferentes (por tanto, no coherentes). Este caso es habitual en acústica al afinar instrumentos.

### 2.1 Expresión Matemática de los Batidos
Consideramos dos ondas en una posición dada $x = 0$:
$$F_1 = A \sin(\omega_1 t)$$
$$F_2 = A \sin(\omega_2 t)$$

Sumando ambas señales:
$$F_T = 2A \cos\left(\frac{\omega_1 - \omega_2}{2} t\right) \sin\left(\frac{\omega_1 + \omega_2}{2} t\right)$$

Definiendo la frecuencia promedio y la diferencia de frecuencias:
$$\omega_{prom} = \frac{\omega_1 + \omega_2}{2}, \quad \Delta\omega = \omega_1 - \omega_2$$
$$F_T(t) = \left[ 2A \cos\left(\frac{\Delta\omega}{2} t\right) \right] \sin(\omega_{prom} t)$$

La respuesta física consiste en una portadora rápida que oscila a la frecuencia promedio $\omega_{prom}$, cuya amplitud está modulada lentamente en el tiempo por el término envolvente:
$$A_{env}(t) = 2A \cos\left(\frac{\Delta\omega}{2} t\right)$$

### 2.2 Frecuencia de Batido
Un "batido" u oído detecta los máximos de volumen acústico (que ocurren tanto en los picos positivos como negativos de la función coseno). Por lo tanto, el periodo de modulación es la mitad del periodo del coseno:
$$T_B = \frac{T_{cos}}{2} \implies f_B = 2 f_{cos} = 2 \left( \frac{f_1 - f_2}{2} \right) = |f_1 - f_2|$$

La **frecuencia de batido** ($f_B$) es igual a la diferencia absoluta de las frecuencias de las dos ondas originales.


```python
# Simulación de Batidos Acústicos
f1 = 220.0   # Frecuencia de la primera onda (Hz - Nota La 3)
f2 = 224.0   # Frecuencia de la segunda onda (Hz - Ligeramente desafinado)
A = 1.0

fs = 2000.0  # Frecuencia de muestreo (Hz)
t = np.linspace(0, 1.5, int(fs * 1.5))

# Ondas individuales
y1 = A * np.sin(2 * np.pi * f1 * t)
y2 = A * np.sin(2 * np.pi * f2 * t)

# Onda resultante
y_result = y1 + y2

# Envolvente analítica del batido
f_beat = np.abs(f1 - f2)
envelope = 2 * A * np.cos(np.pi * f_beat * t)

# Graficación
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(t, y_result, 'b-', label='Señal Resultante $y_1(t) + y_2(t)$', linewidth=1.5)
ax.plot(t, envelope, 'r--', label='Envolvente de modulación $A_{env}(t)$', linewidth=2)
ax.plot(t, -envelope, 'r--', linewidth=2)

ax.set_title(f'Fenómeno de Batidos Acústicos ($f_1 = {f1}$ Hz, $f_2 = {f2}$ Hz $\\\\implies f_{{batido}} = {f_beat}$ Hz)', fontsize=13, fontweight='bold')
ax.set_xlabel('Tiempo $t$ (s)', fontsize=11)
ax.set_ylabel('Amplitud', fontsize=11)
ax.grid(True, linestyle=':', alpha=0.6)
ax.legend(frameon=True, loc='upper right')

plt.tight_layout()
plt.show()
```

## 3. Ondas Estacionarias y Energía en Cuerdas

### 3.1 Energía Transmitida por una Onda en una Cuerda
Para calcular el transporte de energía a lo largo de una cuerda elástica de densidad lineal $\mu$, analizamos un elemento diferencial de longitud $dx$ y masa $dm = \mu dx$.
* **Energía Cinética ($dK$):** Depende del movimiento transversal del elemento ($v_y = \partial y / \partial t$):
  $$dK = \frac{1}{2} dm v_y^2 = \frac{1}{2} (\mu dx) \left(\frac{\partial y}{\partial t}\right)^2$$

Si la onda es armónica viajera ($y = A \sin(kx - \omega t)$), la velocidad transversal es $v_y = -A\omega \cos(kx - \omega t)$:
$$dK = \frac{1}{2} \mu A^2 \omega^2 \cos^2(kx - \omega t) dx$$

Integrando a lo largo de una longitud de onda completa $\lambda$ (período espacial):
$$K_\lambda = \int_0^\lambda \frac{1}{2} \mu A^2 \omega^2 \cos^2(kx - \omega t) dx = \frac{1}{4} \mu A^2 \omega^2 \lambda$$

* **Energía Potencial ($dU$):** Asociada al trabajo realizado para estirar el elemento de cuerda elásticamente:
  $$dU = \frac{1}{2} dm \omega^2 y^2 = \frac{1}{2} (\mu dx) \omega^2 A^2 \sin^2(kx - \omega t)$$
  
Integrando sobre una longitud de onda $\lambda$:
$$U_\lambda = \frac{1}{4} \mu A^2 \omega^2 \lambda$$

* **Energía Mecánica Total ($E_\lambda$):** La suma de ambas:
  $$E_\lambda = K_\lambda + U_\lambda = \frac{1}{2} \mu A^2 \omega^2 \lambda$$

* **Potencia Media ($P$):** Energía transmitida por unidad de periodo temporal $T$:
  $$P = \frac{E_\lambda}{T} = \frac{1}{2} \mu A^2 \omega^2 \left( \frac{\lambda}{T} \right) = \frac{1}{2} \mu A^2 \omega^2 v$$

### 3.2 Velocidad de una Onda en una Cuerda (Derivación por Dinámica)
Consideramos un pequeño elemento de cuerda de longitud $\Delta x$ bajo una tensión constante $T$. Si la perturbación desplaza verticalmente el elemento, el ángulo de inclinación respecto a la horizontal cambia en los extremos ($x$ y $x+\Delta x$):
* Fuerza vertical izquierda: $F_{1y} = -T \sin\alpha \approx -T \tan\alpha = -T \left.\frac{\partial y}{\partial x}\right|_{x}$
* Fuerza vertical derecha: $F_{2y} = T \sin\beta \approx T \tan\beta = T \left.\frac{\partial y}{\partial x}\right|_{x+\Delta x}$

Fuerza neta transversal vertical:
$$F_{neta, y} = T \left[ \left.\frac{\partial y}{\partial x}\right|_{x+\Delta x} - \left.\frac{\partial y}{\partial x}\right|_{x} \right]$$

Por la segunda ley de Newton:
$$F_{neta, y} = dm \cdot a_y = (\mu \Delta x) \frac{\partial^2 y}{\partial t^2}$$
$$T \left[ \frac{\left.\frac{\partial y}{\partial x}\right|_{x+\Delta x} - \left.\frac{\partial y}{\partial x}\right|_{x}}{\Delta x} \right] = \mu \frac{\partial^2 y}{\partial t^2}$$

Tomando el límite continuo $\Delta x \to 0$:
$$T \frac{\partial^2 y}{\partial x^2} = \mu \frac{\partial^2 y}{\partial t^2} \implies \frac{\partial^2 y}{\partial x^2} = \frac{\mu}{T} \frac{\partial^2 y}{\partial t^2}$$

Comparando con la ecuación estándar de onda:
$$\frac{1}{v^2} = \frac{\mu}{T} \implies v = \sqrt{\frac{T}{\mu}}$$


```python
# Simulación espacial de los modos normales de vibración (Armónicas)
# Comparación: Cuerda Fija-Fija vs Cuerda Fija-Libre

L = 1.0  # Longitud de la cuerda (m)
x_plot = np.linspace(0, L, 500)

fig, axes = plt.subplots(3, 2, figsize=(14, 9), sharex=True)

# Modos n = 1, 2, 3
for idx, n in enumerate([1, 2, 3]):
    # 1. Cuerda Fija-Fija (Nodos en ambos extremos): y(0)=0, y(L)=0
    # lambda = 2L/n
    y_fixed = np.sin(n * np.pi * x_plot / L)
    
    axes[idx, 0].plot(x_plot, y_fixed, 'b-', linewidth=2, label=f'Modo n = {n}')
    axes[idx, 0].plot(x_plot, -y_fixed, 'b--', alpha=0.4)
    # Dibujar Nodos (Círculos rojos) y Antinodos (Estrellas verdes)
    nodos_fixed = np.linspace(0, L, n+1)
    axes[idx, 0].scatter(nodos_fixed, np.zeros_like(nodos_fixed), color='red', s=60, zorder=3, label='Nodos')
    axes[idx, 0].set_title(f'Fija-Fija: Armónica {n}', fontsize=11, fontweight='bold')
    axes[idx, 0].grid(True, linestyle=':', alpha=0.6)
    axes[idx, 0].legend(loc='upper right', fontsize=8)
    
    # 2. Cuerda Fija-Libre (Nodo en x=0, Antinodo en x=L): y(0)=0, dy/dx(L)=0
    # lambda = 4L/(2n-1)
    m = 2 * n - 1
    y_free = np.sin(m * np.pi * x_plot / (2 * L))
    
    axes[idx, 1].plot(x_plot, y_free, 'g-', linewidth=2, label=f'Modo n = {n} (Armónica {m})')
    axes[idx, 1].plot(x_plot, -y_free, 'g--', alpha=0.4)
    # Calcular nodos para Fija-Libre: sin(k x) = 0 => x = k_i * lambda / 2
    nodos_free = np.array([i * (2 * L / m) for i in range(n)])
    axes[idx, 1].scatter(nodos_free, np.zeros_like(nodos_free), color='red', s=60, zorder=3, label='Nodos')
    # Antinodo al final
    axes[idx, 1].scatter([L], [y_free[-1]], color='magenta', marker='*', s=120, zorder=3, label='Extremo Libre')
    axes[idx, 1].set_title(f'Fija-Libre: Armónica {m}', fontsize=11, fontweight='bold')
    axes[idx, 1].grid(True, linestyle=':', alpha=0.6)
    axes[idx, 1].legend(loc='upper right', fontsize=8)

axes[-1, 0].set_xlabel('Posición x (m)', fontsize=11)
axes[-1, 1].set_xlabel('Posición x (m)', fontsize=11)
plt.suptitle('Comparación de Modos Normales en Cuerdas con Distintas Condiciones de Contorno', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. Ondas Sonoras Estacionarias y el Tubo de Rubens

El sonido se propaga en tubos llenos de fluido (como columnas de aire en instrumentos de viento) en forma de ondas longitudinales de presión acústica $\Delta p(x, t)$. Al superponerse ondas viajando en sentidos opuestos, se forman resonancias estacionarias.

### 4.1 Ondas Sonoras en Tubo Abierto en Ambos Extremos
En los extremos abiertos de un tubo, el aire tiene plena libertad para moverse (antinodo de desplazamiento molecular). En consecuencia, la variación de presión respecto a la presión atmosférica exterior se reduce a cero, forzando **nodos de presión** en los extremos:
$$\Delta p(0, t) = \Delta p(L, t) = 0$$

La longitud de onda y frecuencias permitidas son análogas a las de una cuerda fija-fija:
$$\lambda_n = \frac{2L}{n}, \quad f_n = n \left(\frac{v}{2L}\right), \quad n \in \{1, 2, 3, \dots\}$$
Donde $v$ representa la velocidad del sonido en el aire ($\sim 343$ m/s a $20^\circ$C).

### 4.2 Ondas Sonoras en Tubo Cerrado en Un Extremo (Cerrado-Abierto)
En el extremo cerrado, las moléculas de aire no pueden desplazarse debido a la pared física (nodo de desplazamiento), lo que produce un **antinodo de presión**. En el extremo abierto se mantiene el nodo de presión:
* Extremo cerrado ($x=0$): Antinodo de presión ($\partial \Delta p / \partial x = 0$).
* Extremo abierto ($x=L$): Nodo de presión ($\Delta p = 0$).

Esto fuerza la cuantización de modos impares:
$$\lambda_n = \frac{4L}{2n - 1}, \quad f_n = (2n - 1) \left(\frac{v}{4L}\right), \quad n \in \{1, 2, 3, \dots\}$$

### 4.3 El Tubo de Rubens
El **Tubo de Rubens** es un instrumento experimental desarrollado por Heinrich Rubens a inicios del siglo XX para visualizar la distribución espacial de la presión acústica en una onda estacionaria.
Consiste en un tubo metálico perforado con una fila de pequeños agujeros en su parte superior, sellado por un extremo con una membrana conectada a un altavoz y por el otro alimentado con gas inflamable (propano). Al encender el gas, las llamas escapan por los agujeros.
* **Física de Flujo:** El flujo de gas es directamente proporcional a la diferencia de presión local dentro del tubo y la presión atmosférica exterior.
* **Comportamiento en Resonancia:** En los **nodos de presión**, la presión del gas se mantiene estable y alta, lo que resulta en **llamas más altas**. En los **antinodos de presión** (donde la presión oscila rápidamente alcanzando valores mínimos), el flujo promedio de gas disminuye debido a efectos de succión dinámica, resultando en **llamas más bajas**. El perfil de altura de las llamas dibuja de forma directa el patrón espacial de la onda acústica estacionaria.


```python
# Simulación Física de las Llamas en un Tubo de Rubens
# Modelado del perfil de altura de las llamas h(x) bajo una resonancia acústica

L_tubo = 2.0  # Longitud del tubo de Rubens (m)
x_tubo = np.linspace(0, L_tubo, 200)
v_sound = 343.0

# Frecuencia del altavoz ajustada a la 4ta armónica del tubo abierto (n = 4)
n = 4
f_exc = n * (v_sound / (2 * L_tubo))  # f = 4 * 343 / 4 = 343 Hz
k = n * np.pi / L_tubo

# Distribución espacial de la presión RMS (Root Mean Square) en resonancia
# En los extremos x=0 y x=L hay nodos de presión (tubo abierto en términos de dinámica de presión)
delta_p_rms = np.abs(np.sin(k * x_tubo))

# Altura de las llamas h(x)
# La altura es máxima en los nodos de presión (delta_p_rms = 0) y mínima en los antinodos
h_base = 15.0  # Altura base de la llama sin sonido (cm)
alpha = 7.0    # Coeficiente de modulación acústica
h_flame = h_base - alpha * (delta_p_rms**2)

# Graficación simulada del Tubo de Rubens
fig, ax = plt.subplots(figsize=(12, 5))

# Dibujar el tubo metálico
ax.axhline(0, color='gray', linewidth=8)
ax.fill_between([0, L_tubo], [-1.5, -1.5], [-0.5, -0.5], color='silver', zorder=2)

# Dibujar las llamas individuales (como barras rellenas de color de fuego)
for xi, hi in zip(x_tubo[::2], h_flame[::2]):
    ax.plot([xi, xi], [0, hi], color='orange', linewidth=2, zorder=1)
    ax.scatter(xi, hi, color='red', s=20, zorder=3)  # Punta de la llama

ax.set_title(f'Simulación de Llamas en un Tubo de Rubens en Resonancia ($f = {f_exc:.1f}$ Hz, Modo $n={n}$)', fontsize=13, fontweight='bold')
ax.set_xlabel('Posición a lo largo del tubo x (m)', fontsize=11)
ax.set_ylabel('Altura de la Llama (cm)', fontsize=11)
ax.set_xlim(-0.1, L_tubo + 0.1)
ax.set_ylim(-2.0, h_base + 3.0)
ax.grid(True, linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()
```

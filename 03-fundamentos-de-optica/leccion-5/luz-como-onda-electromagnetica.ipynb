## 1. Repaso de los Fenómenos Electromagnéticos

Históricamente, la electricidad y el magnetismo se estudiaron como disciplinas independientes. No fue sino hasta el siglo XIX cuando experimentos clave demostraron su íntima conexión:
* **William Gilbert (1600):** Introdujo la palabra "electricidad" y estudió el magnetismo natural de la Tierra.
* **Charles Coulomb (fines del S. XVIII):** Formuló matemáticamente la fuerza entre cargas estáticas puntuales, similar a la ley de gravedad cuadrática inversa:
  $$\vec{F} = \frac{Q_1 Q_2}{4\pi \epsilon_0 r^2} \hat{r}$$
  Donde $\epsilon_0 \approx 8.854 \times 10^{-12}\text{ F/m}$ es la permitividad eléctrica del vacío.
* **Hans Christian Oersted (1820):** Descubrió que una corriente eléctrica desvía una aguja magnética, evidenciando que el movimiento de cargas genera campos magnéticos.
* **Michael Faraday:** Formuló el principio de inducción electromagnética, demostrando que un campo magnético variable en el tiempo induce una corriente eléctrica (y por ende, un campo eléctrico).
* **James Clerk Maxwell (1860s):** Unificó formalmente ambas disciplinas en una sola teoría electromagnética mediante cuatro ecuaciones fundamentales.

### 1.1 Electrostática y Campo Eléctrico
Una carga $Q$ modifica el espacio circundante creando un **campo eléctrico** $\vec{E}$. El campo es la fuerza por unidad de carga de prueba $q$:
$$\vec{E} = \frac{\vec{F}}{q} = \frac{Q}{4\pi \epsilon_0 r^2} \hat{r}$$
El trabajo $W$ necesario para mover una carga entre dos puntos en un campo eléctrico es independiente de la trayectoria (fuerza conservativa). Se define el **potencial eléctrico** $V$ como la energía potencial por unidad de carga:
$$V = \frac{E_p}{q} = \frac{Q}{4\pi \epsilon_0 r}$$
La diferencia de potencial se mide en Voltios ($1\text{ V} = 1\text{ J/C}$). Esto permite definir el **electronvoltio (eV)** como una unidad de energía a escala subatómica:
$$1\text{ eV} = 1.602 \times 10^{-19}\text{ J}$$

Las **líneas de campo** permiten visualizar este campo vectorial. Salen de cargas positivas y entran en negativas. Su densidad es proporcional a la magnitud del campo. El **flujo eléctrico** $\Phi_E$ a través de una superficie cerrada está dado por la **ley de Gauss**:
$$\Phi_E = \oint \vec{E} \cdot d\vec{S} = \frac{Q_{\text{enc}}}{\epsilon_0}$$

### 1.2 Magnetostática y Campo Magnético
Una carga $q$ que se mueve con velocidad $\vec{v}$ en un campo magnético $\vec{B}$ experimenta la **fuerza de Lorentz**:
$$\vec{F} = q \vec{v} \times \vec{B}$$
La dirección de la fuerza se determina mediante la **regla de la mano derecha**. Debido al producto cruz, la fuerza es siempre perpendicular a la velocidad y al campo magnético, por lo que no realiza trabajo sobre la partícula.

La **ley de Biot-Savart** permite calcular el campo magnético inducido por corrientes eléctricas estacionarias:
$$\vec{B} = \frac{\mu_0 I}{4\pi} \oint \frac{d\vec{l} \times \hat{r}}{r^2}$$
Donde $\mu_0 = 4\pi \times 10^{-7}\text{ H/m}$ es la permeabilidad magnética del vacío.

A diferencia de las cargas eléctricas, las líneas de campo magnético siempre forman bucles cerrados. No existen los **monopolos magnéticos** aislados, lo cual se describe formalmente mediante la ley de Gauss para el magnetismo:
$$\oint \vec{B} \cdot d\vec{S} = 0$$
Por último, la **ley de Ampere** relaciona la circulación de $\vec{B}$ a lo largo de una trayectoria cerrada con la corriente neta enlazada:
$$\oint \vec{B} \cdot d\vec{l} = \mu_0 I$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar estilo de visualización
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Definir grilla de coordenadas
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)

# Posición de las cargas del dipolo (+q en x=-1, -q en x=1)
q = 1.0
pos_pos = np.array([-1.0, 0.0])
pos_neg = np.array([1.0, 0.0])

# Calcular distancia a cada carga
r_pos = np.sqrt((X - pos_pos[0])**2 + (Y - pos_pos[1])**2)
r_neg = np.sqrt((X - pos_neg[0])**2 + (Y - pos_neg[1])**2)

# Evitar divisiones por cero en las posiciones exactas de las cargas
r_pos[r_pos < 0.1] = 0.1
r_neg[r_neg < 0.1] = 0.1

# Calcular potencial electrostático total V(x, y)
V = q / (4 * np.pi * r_pos) - q / (4 * np.pi * r_neg)

# Calcular componentes del campo eléctrico E = -grad(V)
Ex = q * (X - pos_pos[0]) / (4 * np.pi * r_pos**3) - q * (X - pos_neg[0]) / (4 * np.pi * r_neg**3)
Ey = q * (Y - pos_pos[1]) / (4 * np.pi * r_pos**3) - q * (Y - pos_neg[1]) / (4 * np.pi * r_neg**3)

# Graficar líneas de campo y superficies equipotenciales
plt.figure(figsize=(10, 8))

# Graficar equipotenciales (contornos)
levels = np.linspace(-0.5, 0.5, 21)
contour = plt.contour(X, Y, V, levels=levels, cmap='RdBu', alpha=0.6, linewidths=1.2)
plt.colorbar(contour, label='Potencial Eléctrico $V$ (unidades arbitrarias)')

# Graficar líneas de campo con vectores de flujo (streamplot)
E_magnitude = np.sqrt(Ex**2 + Ey**2)
plt.streamplot(X, Y, Ex, Ey, color=E_magnitude, cmap='viridis', density=1.5, linewidth=1.0, arrowsize=1.2)

# Graficar las cargas
plt.scatter(pos_pos[0], pos_pos[1], color='red', s=150, zorder=5, label='Carga Positiva (+q)')
plt.text(pos_pos[0]-0.15, pos_pos[1]-0.3, '+q', color='red', fontweight='bold', fontsize=12)
plt.scatter(pos_neg[0], pos_neg[1], color='blue', s=150, zorder=5, label='Carga Negativa (-q)')
plt.text(pos_neg[0]-0.15, pos_neg[1]-0.3, '-q', color='blue', fontweight='bold', fontsize=12)

plt.title('Líneas de Campo Eléctrico y Superficies Equipotenciales de un Dipolo', fontsize=13, fontweight='bold')
plt.xlabel('Eje X (m)', fontsize=11)
plt.ylabel('Eje Y (m)', fontsize=11)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper right')
plt.gca().set_aspect('equal')
plt.tight_layout()
plt.show()
```

## 2. Las Ecuaciones de Maxwell y la Ecuación de Onda

La unificación de Maxwell se basa en la forma diferencial de las cuatro leyes fundamentales en el vacío (donde no existen cargas estáticas ni corrientes libres, $\rho = 0$ y $\vec{J} = 0$):

1. **Ley de Gauss:**
   $$\nabla \cdot \vec{E} = 0$$
2. **Ley de Gauss del Magnetismo:**
   $$\nabla \cdot \vec{B} = 0$$
3. **Ley de Faraday-Maxwell:**
   $$\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}$$
4. **Ley de Ampere-Maxwell:**
   $$\nabla \times \vec{B} = \mu_0 \epsilon_0 \frac{\partial \vec{E}}{\partial t}$$

### 2.1 Derivación Matemática de la Ecuación de Onda
Para obtener una ecuación diferencial en términos de un solo campo, tomamos el rotacional ($\nabla \times$) de la ley de Faraday-Maxwell (Ecuación 3):
$$\nabla \times (\nabla \times \vec{E}) = \nabla \times \left( -\frac{\partial \vec{B}}{\partial t} \right) = -\frac{\partial}{\partial t} (\nabla \times \vec{B})$$

Aplicando la identidad del análisis vectorial para el doble rotacional de un campo vectorial:
$$\nabla \times (\nabla \times \vec{E}) = \nabla(\nabla \cdot \vec{E}) - \nabla^2 \vec{E}$$
Dado que en vacío se cumple $\nabla \cdot \vec{E} = 0$ (Ecuación 1), el lado izquierdo se reduce a:
$$\nabla \times (\nabla \times \vec{E}) = -\nabla^2 \vec{E}$$

Sustituyendo el rotacional de $\vec{B}$ en el lado derecho mediante la ley de Ampere-Maxwell (Ecuación 4):
$$-\nabla^2 \vec{E} = -\frac{\partial}{\partial t} \left( \mu_0 \epsilon_0 \frac{\partial \vec{E}}{\partial t} \right) \implies -\nabla^2 \vec{E} = -\mu_0 \epsilon_0 \frac{\partial^2 \vec{E}}{\partial t^2}$$
Multiplicando por $-1$ obtenemos la **ecuación tridimensional de onda para el campo eléctrico**:
$$\nabla^2 \vec{E} = \mu_0 \epsilon_0 \frac{\partial^2 \vec{E}}{\partial t^2}$$

Siguiendo un proceso análogo (tomando el rotacional de la Ecuación 4 y sustituyendo la ley de Faraday), deducimos la **ecuación de onda para el campo magnético**:
$$\nabla^2 \vec{B} = \mu_0 \epsilon_0 \frac{\partial^2 \vec{B}}{\partial t^2}$$

### 2.2 Velocidad de la Luz en el Vacío
Al comparar estas expresiones con la ecuación diferencial general de onda ($\nabla^2 \Psi = \frac{1}{v^2} \frac{\partial^2 \Psi}{\partial t^2}$), se observa directamente que la velocidad de propagación $c$ de estas ondas en el vacío está dada por:
$$c = \frac{1}{\sqrt{\mu_0 \epsilon_0}}$$
Introduciendo los valores de las constantes físicas del vacío:
* $\mu_0 = 4\pi \times 10^{-7}\text{ H/m}$ (permeabilidad magnética)
* $\epsilon_0 \approx 8.854187 \times 10^{-12}\text{ F/m}$ (permitividad eléctrica)
$$c = \frac{1}{\sqrt{(4\pi \times 10^{-7}) \times (8.854187 \times 10^{-12})}} \approx 299\,792\,458\text{ m/s}$$
Este resultado histórico demostró teóricamente que **la luz es una onda electromagnética**, unificando la óptica y el electromagnetismo.

En un medio material con permeabilidad $\mu = \mu_r \mu_0$ y permitividad $\epsilon = \epsilon_r \epsilon_0$, la velocidad disminuye a:
$$v = \frac{1}{\sqrt{\mu \epsilon}} = \frac{c}{\sqrt{\mu_r \epsilon_r}} = \frac{c}{n}$$
Donde $n = \sqrt{\mu_r \epsilon_r}$ es el **índice de refracción** del medio (típicamente $n \approx \sqrt{\epsilon_r}$ ya que la mayoría de medios dieléctricos tienen $\mu_r \approx 1$).


```python
import sympy as sp

# Definir variables simbólicas
x, y, z, t = sp.symbols('x y z t', real=True)
E0, B0, k, omega, mu0, eps0 = sp.symbols('E0 B0 k omega mu0 eps0', positive=True)

# Definir campos de una onda plana que se propaga en el eje x con polarización en y (E) y z (B)
# E(x, t) = E0 * cos(k*x - omega*t) * j_hat
# B(x, t) = B0 * cos(k*x - omega*t) * k_hat
Ex = 0
Ey = E0 * sp.cos(k*x - omega*t)
Ez = 0

Bx = 0
By = 0
Bz = B0 * sp.cos(k*x - omega*t)

# 1. Verificar divergencia de E (nabla . E = dEx/dx + dEy/dy + dEz/dz)
div_E = sp.diff(Ex, x) + sp.diff(Ey, y) + sp.diff(Ez, z)

# 2. Verificar divergencia de B (nabla . B = dBx/dx + dBy/dy + dBz/dz)
div_B = sp.diff(Bx, x) + sp.diff(By, y) + sp.diff(Bz, z)

# 3. Verificar ley de Faraday: curl(E) = -dB/dt
# curl(E)_x = dEz/dy - dEy/dz = 0
# curl(E)_y = dEx/dz - dEz/dx = 0
# curl(E)_z = dEy/dx - dEx/dy
curl_E_z = sp.diff(Ey, x) - sp.diff(Ex, y)
neg_dB_dt_z = -sp.diff(Bz, t)
faraday_check = sp.simplify(curl_E_z - neg_dB_dt_z)

# Para que se cumpla Faraday, faraday_check debe ser cero, lo cual exige una relación entre E0 y B0
# curl_E_z = -k*E0*sin(...) y -dB_z/dt = -omega*B0*sin(...)
# Por lo tanto: k*E0 = omega*B0 => E0 = (omega/k)*B0 = c*B0

# 4. Verificar ley de Ampere-Maxwell en vacío: curl(B) = mu0 * eps0 * dE/dt
# curl(B)_y = dBx/dz - dBz/dx = -dBz/dx = k*B0*sin(...)
# mu0 * eps0 * dEy/dt = mu0 * eps0 * omega * E0 * sin(...)
curl_B_y = sp.diff(Bx, z) - sp.diff(Bz, x)
ampere_rhs_y = mu0 * eps0 * sp.diff(Ey, t)
ampere_check = sp.simplify(curl_B_y - ampere_rhs_y)

print('--- VERIFICACIÓN DE ECUACIONES DE MAXWELL CON SYMPY ---')
print(f'1. Divergencia de E: {div_E} (Debe ser 0) -> Cumplido')
print(f'2. Divergencia de B: {div_B} (Debe ser 0) -> Cumplido')
print(f'3. Términos de Faraday residual (curl_E - (-dB/dt)): {faraday_check}')
print(f'   Esto se anula si: E0 = (omega/k) * B0 = c * B0')
print(f'4. Términos de Ampere residual (curl_B - mu0*eps0*dE/dt): {ampere_check}')
print(f'   Sustituyendo omega/k = c y c = 1/sqrt(mu0*eps0), se anula exactamente.')
```

## 3. Propiedades y Polarización de las Ondas Electromagnéticas

Las soluciones de las ecuaciones de onda plana revelan propiedades fundamentales de la luz:
1. **Carácter Transversal:** Los campos eléctrico $\vec{E}$ y magnético $\vec{B}$ son perpendiculares a la dirección de propagación. La onda es estrictamente transversal.
2. **Ortogonalidad entre Campos:** $\vec{E}$ y $\vec{B}$ son perpendiculares entre sí ($\vec{E} \cdot \vec{B} = 0$).
3. **Fase:** Ambos campos oscilan en fase (alcanzan sus máximos y mínimos simultáneamente en el espacio y en el tiempo).
4. **Relación de Amplitudes:** La amplitud del campo eléctrico es directamente proporcional a la del campo magnético:
   $$E_0 = c \cdot B_0$$
5. **Dualidad Onda-Partícula:** A nivel cuántico, la energía de la radiación se transporta en paquetes discretos llamados **fotones**. La energía de un fotón es directamente proporcional a la frecuencia $f$ de la onda (relación de Planck-Einstein):
   $$E = h \cdot f = \frac{h \cdot c}{\lambda}$$
   Donde $h \approx 6.626 \times 10^{-34}\text{ J s}$ es la constante de Planck.

### 3.1 Polarización de la Luz
La polarización es una propiedad exclusiva de las ondas transversales que define la orientación espacial de las oscilaciones del campo eléctrico $\vec{E}$ en el plano perpendicular a la propagación. Se clasifican en:

* **Polarización Lineal:** El campo $\vec{E}$ oscila a lo largo de una línea recta en un plano fijo. Puede ser lineal horizontal, vertical o diagonal, dependiendo de la relación de componentes del campo.
* **Polarización Circular:** La magnitud del campo $\vec{E}$ permanece constante, pero su dirección rota uniformemente describiendo una trayectoria circular en el plano transversal. Se origina cuando dos componentes ortogonales del campo tienen la misma amplitud pero están desfasadas exactamente $90^\circ$ ($\delta = \pm \pi/2$):
  $$\vec{E}(z, t) = E_0 \cos(kz - \omega t) \hat{i} \pm E_0 \sin(kz - \omega t) \hat{j}$$
  Dependiendo del sentido de rotación se clasifica en circular izquierda o derecha.
* **Polarización Elíptica:** Es el estado de polarización más general, donde la magnitud y la dirección del campo $\vec{E}$ cambian, trazando una elipse en el plano transversal. Ocurre cuando las componentes ortogonales difieren en amplitud y/o poseen una diferencia de fase cualquiera $\delta \neq n\pi$.


```python
from mpl_toolkits.mplot3d import Axes3D

z_prop = np.linspace(0, 4 * np.pi, 200)  # Eje de propagación Z

# 1. Polarización Lineal Diagonal (En fase: delta = 0, amplitudes iguales)
Ex_lin = np.cos(z_prop)
Ey_lin = np.cos(z_prop)

# 2. Polarización Circular (Desfase delta = pi/2, amplitudes iguales)
Ex_circ = np.cos(z_prop)
Ey_circ = np.sin(z_prop)

# 3. Polarización Elíptica (Desfase delta = pi/4, amplitudes distintas)
Ex_elip = 1.0 * np.cos(z_prop)
Ey_elip = 0.6 * np.cos(z_prop + np.pi/4)

fig = plt.figure(figsize=(15, 6))

# Subplot 1: Polarización Lineal
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.plot(z_prop, Ex_lin, Ey_lin, color='blue', linewidth=2.5, label='Línea de campo $\\vec{E}$')
ax1.plot(z_prop, Ex_lin, -1.5 * np.ones_like(z_prop), 'g--', alpha=0.5)  # Proyección XZ
ax1.plot(z_prop, 1.5 * np.ones_like(z_prop), Ey_lin, 'r--', alpha=0.5)   # Proyección YZ
ax1.set_title('Polarización Lineal', fontsize=12, fontweight='bold')
ax1.set_xlabel('Propagación Z')
ax1.set_ylabel('Eje X')
ax1.set_zlabel('Eje Y')
ax1.set_zlim(-1.2, 1.2)
ax1.view_init(elev=20, azim=-45)

# Subplot 2: Polarización Circular
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.plot(z_prop, Ex_circ, Ey_circ, color='purple', linewidth=2.5)
ax2.plot(z_prop, Ex_circ, -1.5 * np.ones_like(z_prop), 'g--', alpha=0.5)
ax2.plot(z_prop, 1.5 * np.ones_like(z_prop), Ey_circ, 'r--', alpha=0.5)
ax2.set_title('Polarización Circular', fontsize=12, fontweight='bold')
ax2.set_xlabel('Propagación Z')
ax2.set_ylabel('Eje X')
ax2.set_zlabel('Eje Y')
ax2.set_zlim(-1.2, 1.2)
ax2.view_init(elev=20, azim=-45)

# Subplot 3: Polarización Elíptica
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.plot(z_prop, Ex_elip, Ey_elip, color='crimson', linewidth=2.5)
ax3.plot(z_prop, Ex_elip, -1.5 * np.ones_like(z_prop), 'g--', alpha=0.5)
ax3.plot(z_prop, 1.5 * np.ones_like(z_prop), Ey_elip, 'r--', alpha=0.5)
ax3.set_title('Polarización Elíptica', fontsize=12, fontweight='bold')
ax3.set_xlabel('Propagación Z')
ax3.set_ylabel('Eje X')
ax3.set_zlabel('Eje Y')
ax3.set_zlim(-1.2, 1.2)
ax3.view_init(elev=20, azim=-45)

plt.tight_layout()
plt.show()
```

## 4. Energía y Vector de Poynting

Las ondas electromagnéticas transportan energía a través del espacio. Esta energía está almacenada en los propios campos eléctrico y magnético.

### 4.1 Densidad de Energía Electromagnética
La densidad de energía eléctrica $u_E$ y magnética $u_B$ (energía almacenada por unidad de volumen) en el vacío vienen dadas por:
$$u_E = \frac{1}{2} \epsilon_0 E^2, \quad u_B = \frac{1}{2\mu_0} B^2$$
La densidad de energía total $u$ es la suma de ambas:
$$u = u_E + u_B = \frac{1}{2} \epsilon_0 E^2 + \frac{1}{2\mu_0} B^2$$
Sabiendo que para una onda electromagnética plana en vacío se cumple $B = E/c$ y $c^2 = 1/(\mu_0 \epsilon_0)$, se deduce de forma directa que la densidad de energía eléctrica es igual a la magnética ($u_E = u_B$). Por lo tanto:
$$u = \epsilon_0 E^2 = \frac{B^2}{\mu_0}$$

### 4.2 Vector de Poynting (Flujo de Energía)
El transporte de energía se describe mediante el **vector de Poynting** ($\vec{S}$), definido como el flujo de potencia por unidad de área que atraviesa un plano perpendicular a la propagación:
$$\vec{S} = \frac{1}{\mu_0} \vec{E} \times \vec{B}$$
Su dirección apunta hacia la propagación de la onda, y su unidad en el SI es Watts por metro cuadrado ($\text{W/m}^2$). Para una onda plana sinusoidal viajando en el eje $+x$:
$$S(x, t) = c \epsilon_0 E_0^2 \cos^2(kx - \omega t)$$

### 4.3 Intensidad de la Onda
Dado que los campos oscilan a frecuencias extremadamente altas (ej. $\sim 10^{14}\text{ Hz}$ en la luz visible), los detectores físicos no pueden medir la fluctuación instantánea de $S$. En su lugar, se mide la media temporal a lo largo de un ciclo (periodo $T$):
$$I = S_{\text{medio}} = \langle S \rangle = \frac{1}{T} \int_{0}^{T} c \epsilon_0 E_0^2 \cos^2(kx - \omega t) dt$$
Como el promedio de la función $\cos^2(\theta)$ a lo largo de un ciclo es exactamente $1/2$, la **intensidad** $I$ queda expresada como:
$$I = \frac{1}{2} c \epsilon_0 E_0^2 = \frac{c B_0^2}{2\mu_0} = \frac{E_0 B_0}{2\mu_0}$$

### 4.4 Presión de Radiación y Densidad de Momento
Aunque no poseen masa inercial, las ondas electromagnéticas transportan **momento lineal** ($p$). A nivel cuántico, un fotón tiene momento $p = E/c$. A nivel macroscópico, la densidad de momento $dp/dV$ está relacionada con el vector de Poynting:
$$\frac{dp}{dV} = \frac{S}{c^2}$$

Cuando esta radiación choca contra una superficie, transfiere su momento, ejerciendo una fuerza macroscópica llamada **presión de radiación** ($P_{\text{rad}}$):
* **Absorción Total (Cuerpo Negro):** Toda la energía es absorbida:
  $$P_{\text{rad}} = \frac{I}{c}$$
* **Reflexión Total (Espejo Perfecto):** El momento se transfiere en la incidencia y en el rebote (duplicándose la transferencia):
  $$P_{\text{rad}} = \frac{2I}{c}$$


```python
# Graficar los campos acoplados E, B y la magnitud del vector de Poynting S

x_space = np.linspace(0, 3, 500)
E0_val = 1.0
B0_val = 1.0 / 2.0  # Tomamos c = 2.0 para visualización clara
c_test = 2.0
wavelength_test = 1.5
k_test = 2 * np.pi / wavelength_test

# Calcular campos instantáneos en t = 0
E_field = E0_val * np.cos(k_test * x_space)
B_field = B0_val * np.cos(k_test * x_space)

# Magnitud del vector de Poynting proporcional a E * B
S_poynting = E_field * B_field
I_average = 0.5 * E0_val * B0_val  # Intensidad promedio

plt.figure(figsize=(12, 6))
plt.plot(x_space, E_field, 'b-', label='Campo Eléctrico $E_y(x)$', linewidth=2)
plt.plot(x_space, B_field, 'g-.', label='Campo Magnético $B_z(x)$ (amplificado)', linewidth=2)
plt.plot(x_space, S_poynting, 'r-', label='Vector de Poynting $S_x(x)$', linewidth=2.5)
plt.axhline(I_average, color='crimson', linestyle='--', label='Intensidad Media $I = \\langle S \\rangle$', linewidth=1.5)
plt.axhline(0, color='black', linewidth=1, linestyle='-')

plt.title('Distribución Espacial de Campos y Flujo de Energía (Poynting)', fontsize=13, fontweight='bold')
plt.xlabel('Posición de propagación X (m)', fontsize=11)
plt.ylabel('Amplitud de los campos / Energía', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.5)
plt.xlim(0, 3.0)
plt.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

## 5. Ondas Electromagnéticas Estacionarias y Materia

Cuando una onda electromagnética viaja a través de un material dieléctrico, los campos eléctricos oscilantes provocan que los electrones atómicos vibren de forma síncrona. Estos electrones acelerados reemiten una nueva onda electromagnética secundaria con la misma frecuencia pero con un desfase. La interferencia de la onda incidente con la inducida reduce la velocidad neta de propagación en el medio ($v < c$).

### 5.1 Ondas Electromagnéticas Estacionarias
Al incidir una onda electromagnética armónica linealmente polarizada sobre la superficie de un conductor perfecto (como una placa de plata u oro) en $x = 0$, la alta conductividad obliga a que el campo eléctrico tangencial en la superficie sea nulo para cumplir las condiciones de contorno electromagnéticas:
$$E_{\text{tangencial}}(0, t) = 0$$
Esto causa una reflexión de la onda eléctrica con un desfase de $180^\circ$ (cambio de signo en la amplitud). La superposición de la onda incidente y reflejada genera una **onda estacionaria**.

Las expresiones para los campos eléctrico y magnético resultantes de la superposición son:
$$E_y(x, t) = E_0 \cos(kx + \omega t) - E_0 \cos(-kx + \omega t) = -2 E_0 \sin(kx) \sin(\omega t)$$
$$B_z(x, t) = B_0 \cos(kx + \omega t) + B_0 \cos(-kx + \omega t) = 2 B_0 \cos(kx) \cos(\omega t)$$

### 5.2 Nodos y Antinodos de los Campos
Analizando los perfiles espaciales de la onda estacionaria:
* **Nodos de Campo Eléctrico:** Puntos donde la amplitud de $\vec{E}$ es siempre cero:
  $$\sin(kx) = 0 \implies kx = m\pi \implies x = m \frac{\lambda}{2}, \quad m \in \{0, 1, 2, \dots\}$$
* **Nodos de Campo Magnético:** Puntos donde la amplitud de $\vec{B}$ es siempre cero:
  $$\cos(kx) = 0 \implies kx = (2m+1)\frac{\pi}{2} \implies x = (2m+1) \frac{\lambda}{4}, \quad m \in \{0, 1, 2, \dots\}$$

Esto demuestra que los **nodos del campo eléctrico coinciden con los antinodos del campo magnético y viceversa**. Los campos se encuentran desfasados espacialmente $90^\circ$ ($\pi/2$ radianes).

Si encerramos las ondas entre dos placas conductoras paralelas separadas por una distancia $L$, forzamos nodos de campo eléctrico en ambos extremos ($x=0$ y $x=L$), cuantizando las frecuencias permitidas en la **cavidad resonante**:
$$\lambda_n = \frac{2L}{n}, \quad f_n = n \left( \frac{c}{2L} \right), \quad n \in \{1, 2, 3, \dots\}$$


```python
# Graficar los perfiles espaciales de los campos E y B en una onda estacionaria

L_cavity = 2.0
x_cav = np.linspace(0, L_cavity, 300)
k_cav = 2 * np.pi / L_cavity  # Modo fundamental n=2 para visualización (dos lóbulos)
n_mode = 2
k_mode = n_mode * np.pi / L_cavity

# Perfiles espaciales (amplitudes máximas en el tiempo)
E_envelope = np.sin(k_mode * x_cav)
B_envelope = np.cos(k_mode * x_cav)

plt.figure(figsize=(12, 5))
plt.plot(x_cav, E_envelope, 'b-', label='Envolvente de Campo Eléctrico $|E_y(x)|$', linewidth=2.5)
plt.plot(x_cav, B_envelope, 'g--', label='Envolvente de Campo Magnético $|B_z(x)|$', linewidth=2)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='gray', linewidth=4, label='Paredes Conductoras')
plt.axvline(L_cavity, color='gray', linewidth=4)

plt.title(f'Ondas Estacionarias Electromagnéticas en una Cavidad ($L = {L_cavity}$ m, Modo $n={n_mode}$)', fontsize=13, fontweight='bold')
plt.xlabel('Posición en la cavidad X (m)', fontsize=11)
plt.ylabel('Amplitud de Campo Normalizada', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.5)
plt.xlim(-0.1, L_cavity + 0.1)
plt.legend(frameon=True, loc='upper right')
plt.show()
```

## 6. El Espectro Electromagnético y la Ley de Wien

El conjunto de todas las ondas electromagnéticas ordenadas por su frecuencia y longitud de onda constituye el **espectro electromagnético**. Las bandas principales del espectro, de menor a mayor energía, son:

| Tipo de Radiación | Longitud de Onda (m) | Frecuencia (Hz) | Características Principales |
| :--- | :---: | :---: | :--- |
| **Radio** | $> 10^{-2}$ | $< 3 \times 10^{10}$ | No ionizante. Usada en telecomunicaciones y transmisiones de radio/TV. |
| **Microondas** | $10^{-2} - 10^{-5}$ | $3 \times 10^{10} - 3 \times 10^{13}$ | Usada en radares, telefonía móvil y hornos microondas (resonancia de moléculas de agua). |
| **Infrarrojo** | $10^{-5} - 7 \times 10^{-7}$ | $3 \times 10^{13} - 4.3 \times 10^{14}$ | Radiación térmica. Emitida por cuerpos calientes a temperaturas ordinarias. |
| **Visible** | $7 \times 10^{-7} - 4 \times 10^{-7}$ | $4.3 \times 10^{14} - 7.5 \times 10^{14}$ | Pequeña banda detectable por el ojo humano. Máxima sensibilidad a $\sim 550\text{ nm}$ (verde). |
| **Ultravioleta** | $4 \times 10^{-7} - 10^{-9}$ | $7.5 \times 10^{14} - 3 \times 10^{17}$ | Radiación ionizante leve. Causa bronceado y daño celular. Usada en esterilización. |
| **Rayos X** | $10^{-9} - 10^{-12}$ | $3 \times 10^{17} - 3 \times 10^{20}$ | Ionizante de gran penetración. Usada en radiografías médicas y análisis cristalográficos. |
| **Rayos Gamma** | $< 10^{-12}$ | $> 3 \times 10^{20}$ | Sumamente energética. De origen nuclear o astrofísico de alta energía. Extremadamente peligrosa para la vida. |

### 6.1 Radiación de Cuerpo Negro y Ley de Wien
Un **cuerpo negro** es un objeto ideal que absorbe toda la radiación electromagnética incidente sobre él y emite radiación térmica en un espectro continuo que depende exclusivamente de su temperatura termodinámica $T$. La radiancia espectral $I(\lambda, T)$ viene dada por la **ley de Planck**:
$$I(\lambda, T) = \frac{2\pi h c^2}{\lambda^5} \frac{1}{e^{\frac{hc}{\lambda k_B T}} - 1}$$
Donde $k_B \approx 1.381 \times 10^{-23}\text{ J/K}$ es la constante de Boltzmann.

A medida que la temperatura aumenta, la radiancia total se incrementa dramáticamente y el pico de emisión espectral se desplaza hacia longitudes de onda más cortas (mayor energía). La longitud de onda del pico de emisión ($\lambda_{\text{pico}}$) está gobernada por la **ley de desplazamiento de Wien**:
$$\lambda_{\text{pico}} \cdot T = b$$
Donde la constante de desplazamiento de Wien es:
$$b \approx 2.898 \times 10^{-3}\text{ m K} = 0.002898\text{ m K}$$
Esta ley explica por qué un metal calentado brilla primero en rojo, luego en naranja y finalmente en un color blanco azulado a temperaturas muy altas, y permite calcular la temperatura de las superficies estelares analizando su luz.


```python
# Graficar las curvas de radiancia espectral de cuerpo negro para varias temperaturas

# Constantes físicas
h = 6.626e-34      # Constante de Planck (J s)
c = 2.998e8        # Velocidad de la luz (m/s)
k_B = 1.381e-23     # Constante de Boltzmann (J/K)
b_wien = 2.898e-3   # Constante de Wien (m K)

def planck_law(lam, T):
    # Evitar desbordamiento en el cálculo exponencial
    exponent = np.clip((h * c) / (lam * k_B * T), None, 100)
    return (2 * np.pi * h * c**2) / (lam**5 * (np.exp(exponent) - 1))

# Rango de longitudes de onda de 100 nm a 2000 nm
wavelengths = np.linspace(100e-9, 2000e-9, 1000)
temperatures = [3000, 4000, 5000]  # Kelvin
colors_plot = ['crimson', 'darkorange', 'gold']

plt.figure(figsize=(11, 6))

# Colorear la franja del espectro visible (380 nm a 750 nm) en el fondo
plt.axvspan(380, 750, color='lightgray', alpha=0.3, label='Espectro Visible (380-750 nm)')

for T, color in zip(temperatures, colors_plot):
    # Calcular radiancia en unidades de kW/(m^2 * nm) para visualización
    radiance = planck_law(wavelengths, T) * 1e-12  # conversión de W/m^3 a W/(m^2*nm) y luego a kW
    plt.plot(wavelengths * 1e9, radiance, color=color, linewidth=2.5, label=f'Cuerpo Negro a {T} K')
    
    # Calcular pico de Wien
    lambda_peak_nm = (b_wien / T) * 1e9
    radiance_peak = planck_law(b_wien / T, T) * 1e-12
    plt.scatter(lambda_peak_nm, radiance_peak, color='black', edgecolor=color, s=80, zorder=5)
    plt.vlines(lambda_peak_nm, 0, radiance_peak, colors='gray', linestyles=':', alpha=0.7)

plt.title('Radiancia Espectral de Cuerpo Negro y Ley de Desplazamiento de Wien', fontsize=13, fontweight='bold')
plt.xlabel('Longitud de Onda $\\lambda$ (nm)', fontsize=11)
plt.ylabel('Radiancia Espectral ($10^6$ W/m$^3$)', fontsize=11)
plt.xlim(100, 2000)
plt.ylim(0, 15.0)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

## 7. Verificación de Resultados Numéricos

En esta sección de verificación validamos analíticamente la deducción de la constante de Wien y la constante de la velocidad de la luz mediante cálculos con bibliotecas científicas.


```python
import scipy.constants as const
from scipy.optimize import minimize_scalar

# 1. Verificar la velocidad de la luz calculada a partir de las constantes fundamentales mu0 y eps0
mu0_scipy = const.mu_0
eps0_scipy = const.epsilon_0
c_calculated = 1.0 / np.sqrt(mu0_scipy * eps0_scipy)
c_scipy = const.c

# Comparar con aserción de coma flotante
assert np.isclose(c_calculated, c_scipy, atol=1e-5)

# 2. Obtener la constante de desplazamiento de Wien b de forma teórica
# Wien se deriva de encontrar el máximo de la ley de Planck. Derivando dI/dlam = 0 y resolviendo
# se obtiene la ecuación trascendente: 5 - x = 5 * e^-x, donde x = hc / (lam * k_B * T).
def transc_equation(x):
    return (5 - x) - 5 * np.exp(-x)

# Encontramos el cero de la ecuación trascendente numéricamente
# Para ello definimos la función al cuadrado para minimizarla
res = minimize_scalar(lambda x: transc_equation(x)**2, bounds=(4.0, 5.0), method='bounded')
x_max = res.x

# La constante de Wien teórica es: b = h * c / (x_max * k_B)
h_val = const.h
c_val = const.c
k_B_val = const.k
b_theoretical = (h_val * c_val) / (x_max * k_B_val)

print('--- VERIFICACIÓN DE CONSTANTES FÍSICAS ---')
print(f'Velocidad de la luz c calculada: {c_calculated:.2f} m/s')
print(f'Velocidad de la luz c estándar: {c_scipy:.2f} m/s')
print(f'Raíz trascendente del pico de Wien x_max: {x_max:.5f}')
print(f'Constante de Wien b teórica calculada: {b_theoretical:.6e} m K')
print(f'Constante de Wien b estándar de SciPy: {const.Wien:.6e} m K')

assert np.isclose(b_theoretical, const.Wien, rtol=1e-4)
print('Validaciones de Constantes: ¡EXITOSA! (Los valores analíticos coinciden bajo tolerancias estrictas)')
```

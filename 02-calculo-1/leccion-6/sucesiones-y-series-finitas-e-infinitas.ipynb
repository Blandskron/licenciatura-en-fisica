## 1. Sucesiones de Números Reales y Funciones

### 1.1 Definición de Sucesión
Una **sucesión infinita** se puede concebir como una lista ordenada de números escritos en un orden definido:
$$a_1, a_2, a_3, \dots, a_n, \dots$$
donde $a_1$ recibe el nombre de primer término, $a_2$ es el segundo término, y $a_n$ es el $n$-ésimo término general. Para sucesiones infinitas, cada término $a_n$ posee un sucesor único $a_{n+1}$.

Formalmente, una sucesión infinita es una **función** cuyo dominio es el conjunto de los enteros positivos $\mathbb{N}$ y cuyo rango es un subconjunto de los números reales $\mathbb{R}$:
$$a: \mathbb{N} \to \mathbb{R} \quad \text{donde} \quad a(n) = a_n$$

El dominio también puede extenderse para incluir el cero u otros enteros iniciales (ej. $\{b_n\}_{n=0}^\infty$). Las sucesiones pueden definirse de dos formas básicas:
1. **Explícita**: Mediante una fórmula directa para el $n$-ésimo término (ej. $a_n = 3n - 2$, para $n \ge 1$).
2. **Recursiva**: Mediante términos iniciales y una fórmula que define $a_n$ a partir de términos anteriores (ej. el mismo patrón: $a_1 = 1$, $a_n = a_{n-1} + 3$, para $n \ge 2$).

### 1.2 Sucesión de Funciones
Siguiendo a Rivera, podemos extender el concepto a **sucesiones de funciones**, donde cada término de la sucesión es una función real con un dominio $D$ y contradominio $X$ comunes:
$$S: \mathbb{N} \to \{f \mid f: D \to X\} \quad \text{donde} \quad S(n) = f_n$$
Se denota por $(f_n)_{n=1}^\infty$ o simplemente $(f_n)$.

### 1.3 Los 4 Ejemplos Clásicos de la Literatura
Estudiamos el comportamiento de los cuatro ejemplos fundamentales del texto de Stewart:
1. **Ejemplo a**: $a_n = \frac{n}{n+1} \implies \left\{ \frac{1}{2}, \frac{2}{3}, \frac{3}{4}, \frac{4}{5}, \dots \right\}$.
2. **Ejemplo b**: $a_n = \frac{(-1)^n(n+1)}{3^n} \implies \left\{ -\frac{2}{3}, \frac{3}{9}, -\frac{4}{27}, \frac{5}{81}, \dots \right\}$.
3. **Ejemplo c**: $a_n = \sqrt{n-3}$ para $n \ge 3 \implies \left\{ 0, 1, \sqrt{2}, \sqrt{3}, \dots \right\}$.
4. **Ejemplo d**: $a_n = \cos\left(\frac{n\pi}{6}\right)$ para $n \ge 0 \implies \left\{ 1, \frac{\sqrt{3}}{2}, \frac{1}{2}, 0, \dots \right\}$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Graficamos los primeros 25 términos de las 4 sucesiones del texto
n_vals = np.arange(1, 26)

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Subplot a: a_n = n / (n+1) -> Converge a 1
y_a = n_vals / (n_vals + 1)
axs[0, 0].stem(n_vals, y_a, basefmt='gray')
axs[0, 0].axhline(1.0, color='red', linestyle='--', label='Límite $L = 1$')
axs[0, 0].set_title('a) $a_n = \\frac{n}{n+1}$ (Monótona Creciente)', fontweight='bold')
axs[0, 0].set_xlabel('$n$')
axs[0, 0].set_ylabel('$a_n$')
axs[0, 0].set_ylim(0, 1.2)
axs[0, 0].legend()

# Subplot b: a_n = (-1)^n * (n+1) / 3^n -> Converge a 0 (Oscilante amortiguada)
y_b = ((-1.0)**n_vals * (n_vals + 1)) / (3.0**n_vals)
axs[0, 1].stem(n_vals, y_b, basefmt='gray')
axs[0, 1].axhline(0.0, color='red', linestyle='--', label='Límite $L = 0$')
axs[0, 1].set_title('b) $a_n = \\frac{(-1)^n(n+1)}{3^n}$ (Alternada Amortiguada)', fontweight='bold')
axs[0, 1].set_xlabel('$n$')
axs[0, 1].set_ylabel('$a_n$')
axs[0, 1].legend()

# Subplot c: a_n = sqrt(n-3) para n >= 3 -> Diverge a +infinito
n_c = np.arange(3, 28)
y_c = np.sqrt(n_c - 3.0)
axs[1, 0].stem(n_c, y_c, basefmt='gray')
axs[1, 0].set_title('c) $a_n = \\sqrt{n-3}$ (Divergente monótona)', fontweight='bold')
axs[1, 0].set_xlabel('$n$')
axs[1, 0].set_ylabel('$a_n$')

# Subplot d: a_n = cos(n*pi/6) para n >= 0 -> Oscilante acotada (sin límite)
n_d = np.arange(0, 25)
y_d = np.cos(n_d * np.pi / 6.0)
axs[1, 1].stem(n_d, y_d, basefmt='gray')
axs[1, 1].set_title('d) $a_n = \\cos\\left(\\frac{n\\pi}{6}\\right)$ (Oscilante Acotada)', fontweight='bold')
axs[1, 1].set_xlabel('$n$')
axs[1, 1].set_ylabel('$a_n$')
axs[1, 1].set_ylim(-1.3, 1.3)

plt.tight_layout()
plt.show()
```

## 2. Límite y Convergencia de Sucesiones

### 2.1 Definición Rigurosa de Límite
Una sucesión $\{a_n\}$ **converge** a un límite real $L$ (lo que se escribe como $\lim_{n \to \infty} a_n = L$) si para cada número real $\epsilon > 0$ existe un número entero positivo correspondiente $N$ tal que para todo entero $n$ se cumple:
$$n \ge N \implies |a_n - L| < \epsilon$$

Si no existe ningún número finito $L$ al cual converja la sucesión, se dice que la sucesión **diverge** (o que es divergente). Las sucesiones divergentes pueden tender a $+\infty$, $-\infty$, o bien ser **sucesiones oscilantes** (como la sucesión $a_n = (-1)^n \cdot n$).

### 2.2 Propiedades de los Límites de Sucesiones
Sean $\{a_n\}$ y $\{b_n\}$ sucesiones convergentes y sea $k$ una constante real. Las leyes algebraicas de los límites ordinarios se mantienen:
1. $\lim_{n \to \infty} k = k$
2. $\lim_{n \to \infty} k a_n = k \lim_{n \to \infty} a_n$
3. $\lim_{n \to \infty} (a_n \pm b_n) = \lim_{n \to \infty} a_n \pm \lim_{n \to \infty} b_n$
4. $\lim_{n \to \infty} (a_n \cdot b_n) = \lim_{n \to \infty} a_n \cdot \lim_{n \to \infty} b_n$
5. $\lim_{n \to \infty} \frac{a_n}{b_n} = \frac{\lim a_n}{\lim b_n}$, siempre que $\lim_{n \to \infty} b_n \neq 0$.

### 2.3 Teorema de Weierstrass (Monotonía y Acotación)
Para fundamentar la convergencia de sucesiones sin calcular explícitamente el límite, se recurre a la estructura del conjunto real:
* **Acotación**: Una sucesión $\{a_n\}$ está acotada superiormente si existe $M \in \mathbb{R}$ tal que $a_n \le M$ para todo $n$; y acotada inferiormente si existe $m$ tal que $a_n \ge m$. Si cumple ambas, es acotada.
  - *Teorema de acotación*: Toda sucesión convergente es acotada. (El recíproco no es cierto, ej. $1, 2, 1, 2, \dots$ es acotada pero divergente).
* **Teorema de Weierstrass**: Toda sucesión monótona (creciente o decreciente) y acotada es convergente.


```python
# Visualización de la Definición Formal de Límite (Epsilon - N)
# Sucesión: a_n = 1 - 1/n. Converge a L = 1.
# Fijamos un epsilon de tolerancia
epsilon = 0.08
L = 1.0

# Condición: |a_n - 1| < epsilon -> |-1/n| < epsilon -> 1/n < epsilon -> n > 1/epsilon
# Para epsilon = 0.08, n > 12.5 -> El umbral es N = 13.
N_threshold = int(np.ceil(1.0 / epsilon))

n_plot = np.arange(1, 40)
a_plot = 1.0 - 1.0 / n_plot

plt.figure(figsize=(10, 6.5))
plt.stem(n_plot, a_plot, basefmt='gray', label='$a_n = 1 - 1/n$')

# Líneas de límite y tolerancia
plt.axhline(L, color='red', linestyle='-', linewidth=2.0, label='Límite $L = 1$')
plt.axhline(L + epsilon, color='coral', linestyle=':', linewidth=1.5, label='Banda de tolerancia $\\epsilon = 0.08$')
plt.axhline(L - epsilon, color='coral', linestyle=':')

# Marcar umbral N
plt.axvline(N_threshold, color='darkgreen', linestyle='--', linewidth=2.0, 
            label=f'Umbral $N = {N_threshold}$ (n >= N => |a_n - L| < epsilon)')
plt.scatter(n_plot[n_plot >= N_threshold], a_plot[n_plot >= N_threshold], color='green', s=45, zorder=5)

plt.title('Demostración Gráfica de la Definición de Límite de una Sucesión', fontweight='bold')
plt.xlabel('Término ($n$)')
plt.ylabel('$a_n$')
plt.xlim(0.5, 30.5)
plt.ylim(0, 1.2)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 3. Series Infinitas y Prueba de Divergencia

### 3.1 Definición de Serie
Dada una sucesión infinita $\{a_n\}_{n=1}^\infty$, obtenemos una expresión formal al sumar sus infinitos términos:
$$a_1 + a_2 + a_3 + \dots + a_n + \dots = \sum_{n=1}^\infty a_n$$
Esta suma de infinitos términos se denomina **serie infinita** (o simplemente serie).

Para definir rigurosamente la suma de una serie, consideramos las **sumas parciales** acumuladas:
$$s_1 = a_1$$
$$s_2 = a_1 + a_2$$
$$s_3 = a_1 + a_2 + a_3$$
$$s_n = \sum_{i=1}^n a_i = a_1 + a_2 + \dots + a_n$$

Estas sumas parciales forman una nueva sucesión $\{s_n\}_{n=1}^\infty$. Si la sucesión $\{s_n\}$ es convergente y su límite es un número real $s$ (es decir, $\lim_{n \to \infty} s_n = s$), se dice que la serie es **convergente** y escribimos:
$$\sum_{n=1}^\infty a_n = s$$
El número $s$ se denomina la **suma de la serie**. Si la sucesión de sumas parciales $\{s_n\}$ diverge, la serie se dice **divergente**.

### 3.2 Ejemplo Fundamental: La Serie Geométrica
Consideremos la serie geométrica de razón $r = 1/2$:
$$\sum_{n=1}^\infty \frac{1}{2^n} = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \dots$$
Las sumas parciales son:
$$s_1 = 0.5, \quad s_2 = 0.75, \quad s_3 = 0.875, \quad \dots, \quad s_n = 1 - \frac{1}{2^n}$$
Tomando el límite:
$$\lim_{n \to \infty} s_n = \lim_{n \to \infty} \left( 1 - \frac{1}{2^n} \right) = 1 \implies \sum_{n=1}^\infty \frac{1}{2^n} = 1$$

### 3.3 Prueba de la Divergencia (Criterio del Término General)
Existe un teorema fundamental que vincula el comportamiento del término general con la convergencia de la serie:

> **Teorema**: Si la serie $\sum_{n=1}^\infty a_n$ es convergente, entonces el límite de su término general es cero:
> $$\lim_{n \to \infty} a_n = 0$$

* **Advertencia**: El recíproco es falso. Que $\lim_{n \to \infty} a_n = 0$ no garantiza la convergencia de la serie. El contraejemplo clásico es la **serie armónica** $\sum_{n=1}^\infty \frac{1}{n}$, la cual diverge a infinito a pesar de que su término general tiende a cero: $\lim_{n \to \infty} \frac{1}{n} = 0$.

> **La Prueba de la Divergencia**: Si $\lim_{n \to \infty} a_n \neq 0$ o si el límite no existe, entonces la serie $\sum_{n=1}^\infty a_n$ es **divergente**.


```python
# Graficación de la Convergencia de las sumas parciales de la Serie Geométrica
n_series = np.arange(1, 16)
a_terms = 1.0 / (2.0**n_series)
# Sumas parciales acumuladas s_n
s_parciales = np.cumsum(a_terms)

plt.figure(figsize=(10, 6.5))
plt.plot(n_series, s_parciales, 'go--', linewidth=2, markersize=8, label='Sumas Parciales $s_n$')
plt.bar(n_series, a_terms, width=0.4, alpha=0.3, color='blue', label='Términos individuales $a_n$')
plt.axhline(1.0, color='red', linestyle='-', linewidth=2.0, label='Suma de la Serie $s = 1.0$')

for idx, val in zip(n_series, s_parciales):
    plt.text(idx, val - 0.08, f'{val:.4f}', ha='center', fontsize=9, color='darkgreen', fontweight='bold')

plt.title('Convergencia de las Sumas Parciales de la Serie Geométrica $\\sum \\frac{1}{2^n}$', fontweight='bold')
plt.xlabel('Índice ($n$)')
plt.ylabel('Suma acumulada / Valor')
plt.xlim(0.5, 12.5)
plt.ylim(0, 1.2)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. El Criterio Integral y la Función Zeta de Riemann

### 4.1 Criterio Integral de Maclaurin-Cauchy
Cuando no es fácil hallar una fórmula explícita para las sumas parciales $s_n$, recurrimos a criterios de convergencia. El primero de ellos compara la serie con una integral impropia:

> **Teorema**: Supóngase que $f$ es una función continua, positiva y decreciente sobre el intervalo $[1, +\infty)$ y sea $a_n = f(n)$. Entonces la serie $\sum_{n=1}^\infty a_n$ es convergente si y solo si la integral impropia es convergente:
> $$\sum_{n=1}^\infty a_n \text{ converge} \iff \int_1^\infty f(x) dx \text{ converge}$$

### 4.2 Demostración Analítica del Ejemplo del Texto
**Ejemplo**: Evaluar la convergencia de la serie:
$$\sum_{n=1}^\infty \frac{1}{n^2 + 1}$$
Definimos la función asociada $f(x) = \frac{1}{x^2 + 1}$. Esta función es continua, positiva y monótonamente decreciente en $[1, +\infty)$. Aplicamos el criterio integral:
$$\int_1^\infty \frac{1}{x^2 + 1} dx = \lim_{t \to \infty} \int_1^t \frac{1}{x^2 + 1} dx = \lim_{t \to \infty} \left[ \arctan(x) \right]_1^t = \lim_{t \to \infty} \left( \arctan(t) - \arctan(1) \right)$$
Dado que $\lim_{t \to \infty} \arctan(t) = \frac{\pi}{2}$ y $\arctan(1) = \frac{\pi}{4}$:
$$\int_1^\infty \frac{1}{x^2 + 1} dx = \frac{\pi}{2} - \frac{\pi}{4} = \frac{\pi}{4}$$
Puesto que la integral impropia converge a un valor finito, **la serie $\sum_{n=1}^\infty \frac{1}{n^2+1}$ es convergente**.

### 4.3 La Serie $p$ y la Función Zeta de Riemann
La serie de términos recíprocos de potencias se denomina **serie $p$**:
$$\sum_{n=1}^\infty \frac{1}{n^p} = 1 + \frac{1}{2^p} + \frac{1}{3^p} + \dots$$
Aplicando el criterio integral de comparación, podemos analizar su convergencia:
* Para $p = 1$ (Serie Armónica): $\int_1^\infty \frac{1}{x} dx = \lim_{t \to \infty} \ln(t) = +\infty \implies$ Divergente.
* Para $p \neq 1$:
  $$\int_1^\infty \frac{1}{x^p} dx = \lim_{t \to \infty} \left[ \frac{x^{-p+1}}{-p+1} \right]_1^t = \lim_{t \to \infty} \left( \frac{t^{1-p}}{1-p} - \frac{1}{1-p} \right)$$
  - Si $p > 1$, entonces $1-p < 0 \implies t^{1-p} \to 0$, por lo que la integral converge a $\frac{1}{p-1}$.
  - Si $p < 1$, entonces $1-p > 0 \implies t^{1-p} \to +\infty$, por lo que la integral diverge.

Por lo tanto, la serie $p$ (que define la función Zeta de Riemann $\zeta(p)$ para valores reales) **converge si $p > 1$ y diverge si $p \le 1$**.


```python
# Ilustración Geométrica de la Prueba Integral para f(x) = 1/x^2 (p-serie con p = 2)
x_int = np.linspace(1.0, 6.0, 300)
y_int = 1.0 / x_int**2

n_boxes = np.arange(1, 6)
y_boxes = 1.0 / n_boxes**2

plt.figure(figsize=(10, 6.5))
plt.plot(x_int, y_int, 'r-', linewidth=2.5, label='Curva continua $f(x) = 1/x^2$')

# Rectángulos de la Suma de Riemann Superior (por la izquierda) para ilustrar acotación
# Área total de rectángulos = 1/1^2 + 1/2^2 + ... = Suma de la serie
plt.bar(n_boxes, y_boxes, width=1.0, align='edge', alpha=0.25, edgecolor='blue', color='skyblue', 
        label='Suma de la Serie $\\sum_{n=1}^5 \\frac{1}{n^2}$')

plt.title('Comparación Geométrica de Criterio Integral: Serie vs Integral Impropia', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.xlim(0.8, 6.2)
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 5. Series Alternadas y el Criterio de Leibniz

### 5.1 Definición de Serie Alternante
Una **serie alternada** es aquella cuyos términos modifican su signo de manera alternada entre valores positivos y negativos. Se representa mediante:
$$\sum_{n=1}^\infty (-1)^{n-1} b_n = b_1 - b_2 + b_3 - b_4 + \dots \quad (b_n > 0)$$
o bien
$$\sum_{n=1}^\infty (-1)^n b_n = -b_1 + b_2 - b_3 + b_4 - \dots \quad (b_n > 0)$$
donde $b_n = |a_n|$ representa la magnitud absoluta del término general.

### 5.2 Criterio de Leibniz (Prueba de la Serie Alternante)
La convergencia de una serie alternante se establece bajo condiciones muy sencillas e intuitivas:

> **Teorema (Leibniz)**: Si la serie alternante cumple las dos condiciones:
> 1. **Decrecimiento**: $b_{n+1} \le b_n$ para toda $n$.
> 2. **Límite nulo**: $\lim_{n \to \infty} b_n = 0$.
> entonces la serie es **convergente**.

### 5.3 Interpretación Geométrica de la Demostración
Siguiendo la visualización clásica (Fig. 7), analizamos el comportamiento de las sumas parciales:
- Partiendo de $s_1 = b_1$.
- Restamos $b_2$ para obtener $s_2 = b_1 - b_2$ (a la izquierda de $s_1$).
- Sumamos $b_3$ para obtener $s_3 = s_2 + b_3$. Como $b_3 < b_2$, el paso es más corto, por lo que $s_3$ queda a la derecha de $s_2$ pero a la izquierda de $s_1$.
- En general, las sumas parciales pares $s_{2n}$ son crecientes y acotadas superiormente por $s_1$, y las impares $s_{2n-1}$ son decrecientes y acotadas inferiormente por $s_2$. Ambas sucesiones se encierran mutuamente oscilando hacia atrás y hacia adelante, convergiendo a un único límite común $s$.


```python
# Simulación de la oscilación de sumas parciales para la Serie Armónica Alternante
# Serie: sum (-1)^(n-1) / n -> converge a ln(2) ~ 0.693147
n_alt = np.arange(1, 21)
b_alt = 1.0 / n_alt
a_alt = (-1.0)**(n_alt - 1) * b_alt
s_alt = np.cumsum(a_alt)

plt.figure(figsize=(11, 7))
plt.plot(n_alt, s_alt, 'o-', color='darkgreen', linewidth=1.5, label='Sumas Parciales $s_n$')

# Dibujar flechas para ilustrar la oscilación "hacia atrás y hacia adelante"
for i in range(1, len(s_alt)):
    color_arrow = 'blue' if a_alt[i] > 0 else 'red'
    plt.annotate('', xy=(n_alt[i], s_alt[i]), xytext=(n_alt[i-1], s_alt[i-1]),
                 arrowprops=dict(arrowstyle="->", color=color_arrow, lw=1.2, alpha=0.7))

plt.axhline(np.log(2.0), color='black', linestyle='--', linewidth=2.0, 
            label='Límite de convergencia $s = \\ln(2) \\approx 0.6931$')

plt.title('Oscilación de Sumas Parciales en el Criterio de Leibniz (Serie Armónica Alternante)', fontweight='bold')
plt.xlabel('Término ($n$)')
plt.ylabel('Suma acumulada ($s_n$)')
plt.xticks(n_alt)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 6. Convergencia Absoluta, Condicional y Criterio del Cociente

### 6.1 Definiciones
* **Convergencia Absoluta**: Una serie $\sum a_n$ es **absolutamente convergente** si la serie de sus valores absolutos es convergente:
  $$\sum_{n=1}^\infty |a_n| \quad \text{es convergente}$$
* **Convergencia Condicional**: Una serie es **condicionalmente convergente** si la serie original converge pero la serie de sus valores absolutos diverge.
  - *Ejemplo*: La serie armónica alternante $\sum_{n=1}^\infty \frac{(-1)^{n-1}}{n}$ converge (por Leibniz), pero la de valores absolutos es la serie armónica simple $\sum \frac{1}{n}$, la cual diverge. Por tanto, es condicionalmente convergente.

### 6.2 Teorema Fundamental de Convergencia Absoluta
> **Teorema**: Si una serie es absolutamente convergente, entonces es convergente.
>
> *Demostración*: Consideremos la desigualdad evidente para cualquier término:
> $$0 \le a_n + |a_n| \le 2|a_n|$$
> Puesto que la serie $\sum |a_n|$ es convergente por hipótesis, la serie $\sum 2|a_n|$ también es convergente (linealidad).
> Por el Criterio de Comparación Directa para series de términos positivos, la serie intermedia $\sum (a_n + |a_n|)$ es convergente.
> Expresando la serie original como diferencia de dos series convergentes:
> $$\sum a_n = \sum (a_n + |a_n|) - \sum |a_n|$$
> Se concluye que la serie original $\sum a_n$ es convergente. Queda demostrado.

### 6.3 Criterio del Cociente (Criterio de D\'Alembert / Ratio Test)
Es el método más potente para series exponenciales y factoriales. Se basa en evaluar el límite de la razón de magnitudes consecutivas:

> **Criterio**: Sea $\sum a_n$ una serie y supóngase el límite:
> $$\rho = \lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right|$$
> - Si **$\rho < 1$**, la serie es **absolutamente convergente** (y por tanto convergente).
> - Si **$\rho > 1$** (o $\rho = \infty$), la serie es **divergente**.
> - Si **$\rho = 1$**, el criterio es **indeterminado** (inconclusivo; la serie puede converger o diverger).


```python
# Análisis del Criterio del Cociente para la serie del texto: sum (n+1) / 2^n
# Término general: a_n = (n+1) / 2^n. Razón: |a_(n+1)/a_n| = (n+2)/2(n+1) -> Converge a rho = 0.5 < 1
n_ratio = np.arange(1, 40)
a_ratio = (n_ratio + 1.0) / (2.0**n_ratio)

# Calculamos las razones consecutivas
razones = a_ratio[1:] / a_ratio[:-1]

print("Razones consecutivas para la serie sum (n+1)/2^n:")
print(f"{'n':<5} | {'a_n':<12} | {'Razón a_(n+1)/a_n':<22}")
print("-" * 45)
for i in range(8):
    print(f"{n_ratio[i]:<5} | {a_ratio[i]:<12.6f} | {razones[i]:<22.6f}")
print("...")
print(f"Última razón calculada (n={n_ratio[-2]}): {razones[-1]:.6f} (Se aproxima a rho = 0.5)")
print("-" * 45)

# Graficar la convergencia de la razón a rho = 0.5
plt.figure(figsize=(10, 6))
plt.plot(n_ratio[:-1], razones, 'bo-', linewidth=2.0, label='Razones consecutivas $\\left|\\frac{a_{n+1}}{a_n}\\right|$')
plt.axhline(0.5, color='red', linestyle='--', linewidth=2.0, label='Límite de la razón $\\rho = 0.5$')
plt.title('Demostración del Criterio del Cociente (Ratio Test)', fontweight='bold')
plt.xlabel('Término ($n$)')
plt.ylabel('Razón de magnitudes')
plt.ylim(0.3, 0.8)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 7. Resumen de la Lección

1. **Sucesión**: Lista ordenada de números reales modelados formalmente como una función discreta sobre los números naturales. Pueden ser convergentes, divergentes u oscilantes.
2. **Teorema de Weierstrass**: Fundamenta que toda sucesión monótona y acotada converge de forma segura en los números reales.
3. **Serie**: Suma infinita de términos cuya convergencia depende del comportamiento límite de su sucesión de sumas parciales $s_n$.
4. **Prueba de Divergencia**: Si el término general no tiende a cero, la serie diverge de forma inmediata.
5. **Criterio Integral**: Permite evaluar series comparando su decaimiento con el de una integral impropia continua, fundamentando la convergencia de la serie $p$ (Función Zeta de Riemann) para $p > 1$.
6. **Criterio de Leibniz**: Establece que las series alternadas convergen si sus magnitudes decrecen de forma monótona hacia cero, oscilando en torno a su límite.
7. **Criterio del Cociente**: Determina la convergencia absoluta analizando la razón de términos consecutivos $\rho$, siendo especialmente útil para factoriales y exponenciales.

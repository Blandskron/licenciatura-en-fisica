## 1. Concepto de Función

### 1.1 Definición Matemática de Función
En matemáticas, de acuerdo con Mathews *et al.*, una **función** $f$ de un conjunto $A$ en un conjunto $B$ (denotada por $f: A \to B$) es una relación que asocia a cada elemento $x \in A$ uno y solo un elemento $y \in B$, denominado la **imagen** de $x$ bajo $f$, lo que se escribe como $y = f(x)$.

Formalmente, deben cumplirse dos condiciones de existencia y unicidad:
1. **Existencia**: Para todo $x \in A$, existe un elemento $y \in B$ tal que $(x, y) \in f$.
2. **Unicidad**: Si $(x, y_1) \in f$ y $(x, y_2) \in f$, entonces $y_1 = y_2$.

* **Dominio ($Dom(f)$)**: El conjunto $A$ de partida que contiene todas las variables independientes $x$ para las cuales la función está definida.
* **Codominio ($Cod(f)$)**: El conjunto $B$ de llegada de la función.
* **Imagen o Rango ($Im(f)$)**: El subconjunto de $B$ formado por todos los valores que efectivamente toma la función:
  $$Im(f) = \{y \in B \mid \exists x \in A, \ f(x) = y\}$$

### 1.2 Clasificación de Funciones
Las funciones se clasifican según su comportamiento de asignación:
* **Función Inyectiva (Uno a Uno)**: Una función es inyectiva si a elementos distintos del dominio corresponden imágenes distintas en el codominio.
  $$\forall x_1, x_2 \in A, \quad x_1 \neq x_2 \implies f(x_1) \neq f(x_2)$$
  Equivalentemente (por contraposición):
  $$\forall x_1, x_2 \in A, \quad f(x_1) = f(x_2) \implies x_1 = x_2$$
* **Función Sobreyectiva (Suryectiva o Sobre)**: Una función es sobreyectiva si su rango es igual a su codominio. Es decir, todo elemento del codominio es imagen de al menos un elemento del dominio.
  $$\forall y \in B, \quad \exists x \in A \text{ tal que } f(x) = y$$
* **Función Biyectiva**: Una función es biyectiva si es simultáneamente **inyectiva** y **sobreyectiva**. Las funciones biyectivas admiten una función inversa única $f^{-1}: B \to A$.

### 1.3 Ceros, Positividad y Negatividad
Para analizar el comportamiento analítico de una función real, se definen los siguientes conjuntos esenciales:
* **Ceros o Raíces**: Valores del dominio donde la función se anula:
  $$Z = \{x \in Dom(f) \mid f(x) = 0\}$$
* **Conjunto de Positividad ($C^+$)**: Intervalos del dominio donde las imágenes son estrictamente mayores que cero:
  $$C^+ = \{x \in Dom(f) \mid f(x) > 0\}$$
* **Conjunto de Negatividad ($C^-$)**: Intervalos del dominio donde las imágenes son estrictamente menores que cero:
  $$C^- = \{x \in Dom(f) \mid f(x) < 0\}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Creamos una figura con dos subplots para contrastar inyectividad y raíces/conjuntos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Función Inyectiva f(x) = exp(x)
x_inj = np.linspace(-2.0, 2.0, 200)
y_inj = np.exp(x_inj)
ax1.plot(x_inj, y_inj, 'b-', linewidth=2.5, label='$f(x) = e^x$')
# Dibujamos líneas horizontales de prueba para ilustrar el test de la recta horizontal
for y_val in [0.5, 2.0, 5.0]:
    ax1.axhline(y_val, color='gray', linestyle='--', alpha=0.5)
    x_val = np.log(y_val)
    ax1.plot(x_val, y_val, 'ro')
ax1.set_title('Función Inyectiva: $f(x) = e^x$\n(Test de la recta horizontal: corta a lo sumo una vez)', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# Subplot 2: Función No Inyectiva f(x) = x^3 - 3x (con raíces y conjuntos C+ / C-)
x_no_inj = np.linspace(-2.2, 2.2, 300)
y_no_inj = x_no_inj**3 - 3 * x_no_inj
ax2.plot(x_no_inj, y_no_inj, 'g-', linewidth=2.5, label='$f(x) = x^3 - 3x$')

# Identificar raíces (x = -sqrt(3), 0, sqrt(3))
raices = [-np.sqrt(3), 0, np.sqrt(3)]
ax2.plot(raices, [0, 0, 0], 'ko', markersize=8, label='Raíces (Ceros)')

# Colorear regiones C+ (positividad) y C- (negatividad)
ax2.fill_between(x_no_inj, y_no_inj, 0, where=(y_no_inj > 0), color='lightblue', alpha=0.3, label='$C^+$ (Positividad)')
ax2.fill_between(x_no_inj, y_no_inj, 0, where=(y_no_inj < 0), color='lightcoral', alpha=0.3, label='$C^-$ (Negatividad)')

ax2.set_title('Función No Inyectiva: $f(x) = x^3 - 3x$\n(Ceros, Positividad $C^+$ y Negatividad $C^-$)', fontweight='bold')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.axhline(0, color='black', linewidth=0.8)
ax2.axvline(0, color='black', linewidth=0.8)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 2. La Función Lineal y sus Aplicaciones Reales

### 2.1 Definición Matemática de Función Lineal
Una **función lineal** es una función polinómica de primer grado de la forma:
$$f: \mathbb{R} \to \mathbb{R} \quad \text{definida por} \quad f(x) = mx + b$$
donde $m$ y $b$ son números reales constantes.

* **La Pendiente ($m$)**: Determina el grado de inclinación de la recta con respecto al eje de las abscisas ($x$). Representa la razón de cambio constante entre las variables:
  $$m = \frac{\Delta y}{\Delta x} = \frac{f(x_2) - f(x_1)}{x_2 - x_1}$$
  - Si $m > 0$, la función es **estrictamente creciente**.
  - Si $m < 0$, la función es **estrictamente decreciente**.
  - Si $m = 0$, la función es **constante** ($f(x) = b$).
* **La Ordenada al Origen ($b$)**: Es el valor donde la recta interseca el eje de las ordenadas ($y$). Ocurre cuando la variable independiente es cero ($f(0) = b$).
* **Proporcionalidad de los Incrementos**: En una función lineal, el incremento de los valores en el codominio es directamente proporcional al incremento en el dominio:
  $$f(x + \Delta x) - f(x) = m \cdot \Delta x$$

### 2.2 Aplicación de las Funciones Reales en el Mundo Real
Las funciones reales permiten modelar fenómenos en diversas disciplinas científicas:
1. **Economía (Ley de Demanda)**: El precio $P$ de una unidad de producto y la cantidad demandada $x$ por los consumidores suelen relacionarse linealmente mediante $P(x) = mx + b$. La pendiente $m$ suele ser negativa, reflejando que a mayor precio, menor demanda.
2. **Medicina y Psicología Cognitiva (Experimento de Stenberg)**: Utilizado para estudiar el tiempo de recuperación de información de la memoria a corto plazo en humanos. La relación entre el tiempo de reacción $T$ (en milisegundos) y el tamaño del conjunto de estímulos memorizados $s$ es lineal:
   $$T(s) = ms + b$$
   donde $b$ representa el tiempo de procesamiento sensorial y motor (ordenada al origen, usualmente cerca de $400\text{ ms}$), y $m$ es el tiempo medio de comparación en la memoria de trabajo por cada elemento adicional (pendiente, típicamente unos $38\text{ ms}$ por elemento).


```python
# Gráfica de aplicaciones reales de la función lineal
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 1. Experimento de Stenberg: T(s) = 38s + 400 (para s = 1 a 6 elementos memorizados)
s_valores = np.arange(1, 7)
T_valores = 38 * s_valores + 400

ax1.plot(s_valores, T_valores, 'o-', color='crimson', linewidth=2.5, markersize=8, label='$T(s) = 38s + 400$')
for s_val, T_val in zip(s_valores, T_valores):
    ax1.text(s_val, T_val - 12, f'{T_val} ms', ha='center', fontweight='bold', color='darkred')
ax1.set_title('Experimento de Stenberg: Tiempo de Reacción vs Memoria', fontweight='bold')
ax1.set_xlabel('Tamaño del conjunto de estímulos ($s$)', fontsize=11)
ax1.set_ylabel('Tiempo de reacción ($T$, ms)', fontsize=11)
ax1.set_xlim(0.5, 6.5)
ax1.set_ylim(350, 680)
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Ley de Demanda: P(x) = -2x + 100 (x: cantidad de unidades, P: precio en pesos)
x_demanda = np.linspace(0, 50, 100)
P_demanda = -2 * x_demanda + 100

ax2.plot(x_demanda, P_demanda, 'b-', linewidth=2.5, label='Curva de Demanda: $P(x) = -2x + 100$')
ax2.fill_between(x_demanda, P_demanda, color='blue', alpha=0.1, label='Región Factible de Compra')
ax2.set_title('Modelo Económico: Ley de Demanda Lineal', fontweight='bold')
ax2.set_xlabel('Cantidad demandada ($x$)', fontsize=11)
ax2.set_ylabel('Precio por unidad ($P$, pesos)', fontsize=11)
ax2.set_xlim(0, 52)
ax2.set_ylim(0, 110)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 3. Concepto de Límite

### 3.1 Definición Intuitiva de Límite
El concepto de **límite** formaliza matemáticamente la noción de aproximación de una función $f(x)$ a un valor de convergencia $L$ a medida que la variable independiente $x$ se aproxima a un punto de acumulación $a$, sin necesidad de que la función esté definida en dicho punto $a$.

Decimos que el límite de $f(x)$ cuando $x$ tiende a $a$ es $L$, y lo escribimos como:
$$\lim_{x \to a} f(x) = L$$
si podemos hacer que los valores de $f(x)$ estén tan cerca de $L$ como queramos, tomando valores de $x$ suficientemente cercanos a $a$ (tanto por la izquierda como por la derecha), pero no iguales a $a$ ($x \neq a$).

### 3.2 Análisis Numérico del Límite de $f(x) = x^2 - x + 2$ en $x \to 2$
Siguiendo el ejemplo clásico de Stewart *et al.*, investigamos el comportamiento de la función cuadrática:
$$f(x) = x^2 - x + 2$$
cuando la variable independiente $x$ toma valores extremadamente cercanos a $2$.
Evaluando $f(x)$ para valores de $x$ a ambos lados de $2$ ($x < 2$ por la izquierda, y $x > 2$ por la derecha), observamos que los valores correspondientes de la función se aproximan a $4$.


```python
# Función cuadrática de prueba
def f_cuad(x):
    return x**2 - x + 2

# 1. Generación de tablas de aproximación numérica lateral
x_izq = np.array([1.9, 1.95, 1.99, 1.995, 1.999, 1.9999])
x_der = np.array([2.1, 2.05, 2.01, 2.005, 2.001, 2.0001])

y_izq = f_cuad(x_izq)
y_der = f_cuad(x_der)

print("Aproximación por la Izquierda (x < 2):")
print(f"{'x':<10} | {'f(x)':<12}")
print("-" * 25)
for xi, yi in zip(x_izq, y_izq):
    print(f"{xi:<10.4f} | {yi:<12.6f}")

print("\nAproximación por la Derecha (x > 2):")
print(f"{'x':<10} | {'f(x)':<12}")
print("-" * 25)
for xi, yi in zip(x_der, y_der):
    print(f"{xi:<10.4f} | {yi:<12.6f}")

# 2. Graficación del límite en x -> 2
x_vals = np.linspace(1.5, 2.5, 300)
y_vals = f_cuad(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2.5, label='$f(x) = x^2 - x + 2$')

# Dibujar punto del límite
plt.plot(2, 4, 'ro', markersize=9, label='Límite $\\lim_{x \\to 2} f(x) = 4$')
plt.axvline(2, color='gray', linestyle=':', alpha=0.8)
plt.axhline(4, color='gray', linestyle=':', alpha=0.8)

plt.title('Aproximación Numérica e Interpretación Gráfica del Límite', fontweight='bold')
plt.xlabel('Variable independiente ($x$)')
plt.ylabel('Función ($f(x)$)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. Cálculo de Límites y Casos de No Existencia

### 4.1 Estimación de un Límite Racional Indeterminado
Consideremos la función racional:
$$f(x) = \frac{x - 1}{x^2 - 1}$$
Esta función no está definida en $x = 1$, puesto que anula el denominador (dando una indeterminación algebraica del tipo $0/0$). No obstante, al calcular el límite mediante factorización y simplificación algebraica:
$$\lim_{x \to 1} \frac{x - 1}{x^2 - 1} = \lim_{x \to 1} \frac{x - 1}{(x-1)(x+1)} = \lim_{x \to 1} \frac{1}{x+1} = \frac{1}{2} = 0.5$$
Aunque la función tiene un "punto vacío" (agujero) en $x = 1$, el límite existe y es igual a $0.5$.

### 4.2 Casos donde el Límite No Existe
Un límite en un punto $a$ no existe si los valores de la función no convergen a un único valor finito. Veremos dos ejemplos clásicos fundamentales:

1. **Discontinuidad por Salto Finito (La Función de Heaviside)**:
   La función de Heaviside $H(t)$ modela la activación instantánea de una corriente eléctrica en el tiempo $t = 0$:
   $$H(t) = \begin{cases} 0 & \text{si } t < 0 \\ 1 & \text{si } t \ge 0 \end{cases}$$
   Al aproximarnos a $t = 0$ por la izquierda, $H(t) \to 0$. Al aproximarnos por la derecha, $H(t) \to 1$. Dado que los límites laterales difieren en valor finito, el límite global no existe.
   
2. **Comportamiento Oscilatorio Infinito (Función Patológica)**:
   Investiguemos la función trigonométrica:
   $$f(x) = \sin\left(\frac{\pi}{x}\right)$$
   Esta función no está definida en $x = 0$. Conforme $x \to 0$, el término $\frac{\pi}{x}$ crece sin límite, lo cual provoca que la función oscile entre $-1$ y $1$ de manera infinitamente densa en cualquier intervalo alrededor de $0$. Dado que las imágenes no se aproximan a ningún número fijo, el límite $\lim_{x \to 0} \sin\left(\frac{\pi}{x}\right)$ no existe.


```python
# Creamos subplots 1x3 para ilustrar cada caso de límite racional y de no existencia
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6.5))

# Subplot 1: f(x) = (x-1)/(x^2-1) con un agujero en x = 1
x_rac = np.linspace(0.2, 1.8, 400)
x_rac = x_rac[x_rac != 1.0] # Evitamos el punto exacto de la indeterminación
y_rac = (x_rac - 1) / (x_rac**2 - 1)

ax1.plot(x_rac, y_rac, 'b-', linewidth=2.5, label='$f(x) = \\frac{x-1}{x^2-1}$')
ax1.plot(1.0, 0.5, 'ro', fillstyle='none', markersize=10, markeredgewidth=2.0, label='Punto vacío en (1, 0.5)')
ax1.set_title('Límite con Indeterminación Evitable', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$f(x)$')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# Subplot 2: Función de Heaviside H(t)
t_izq = np.linspace(-1.5, -0.0001, 100)
t_der = np.linspace(0.0, 1.5, 100)

ax2.plot(t_izq, np.zeros_like(t_izq), 'g-', linewidth=2.5)
ax2.plot(t_der, np.ones_like(t_der), 'g-', linewidth=2.5, label='Función Heaviside $H(t)$')
ax2.plot(0.0, 0.0, 'go', fillstyle='none', markersize=8, markeredgewidth=1.5) # límite izquierdo
ax2.plot(0.0, 1.0, 'go', markersize=8) # límite derecho e inclusión del punto
ax2.set_title('Límite No Existe: Salto Finito ($H(t)$)', fontweight='bold')
ax2.set_xlabel('$t$')
ax2.set_ylabel('$H(t)$')
ax2.legend(loc='lower right')
ax2.grid(True, linestyle='--', alpha=0.7)

# Subplot 3: Función oscilatoria f(x) = sin(pi/x)
x_osc = np.linspace(-0.5, 0.5, 2000)
x_osc = x_osc[x_osc != 0.0]
y_osc = np.sin(np.pi / x_osc)

ax3.plot(x_osc, y_osc, 'r-', linewidth=1.2, label='$\\sin(\\pi / x)$')
ax3.set_title('Límite No Existe: Oscilación Infinita', fontweight='bold')
ax3.set_xlabel('$x$')
ax3.set_ylabel('$f(x)$')
ax3.set_xlim(-0.25, 0.25)
ax3.legend()
ax3.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 5. Definición Formal de Límite (Epsilon-Delta)

### 5.1 Definición Rigurosa de Cauchy-Weierstrass
La noción intuitiva de aproximación se formaliza mediante la **definición $\epsilon-\delta$** (atribuida a Karl Weierstrass e introducida en el texto de Barreno *et al.*):

> **Definición**: Sea $f: A \to \mathbb{R}$ una función real y sea $x_0$ un punto de acumulación del dominio $A$. Se dice que el límite de $f(x)$ cuando $x$ tiende a $x_0$ es $L \in \mathbb{R}$ si y solo si para todo número real $\epsilon > 0$ existe un número real $\delta > 0$ tal que para todo $x \in A$ que cumpla $0 < |x - x_0| < \delta$, se verifica que $|f(x) - L| < \epsilon$.
>
> En símbolos:
> $$\lim_{x \to x_0} f(x) = L \iff \forall \epsilon > 0, \ \exists \delta > 0 \ \text{tal que} \ \left( 0 < |x - x_0| < \delta \implies |f(x) - L| < \epsilon \right)$$

### 5.2 Interpretación Geométrica
Geométricamente, la definición establece que si fijamos una franja horizontal de tolerancia de ancho $2\epsilon$ centrada en el límite $L$ (es decir, el intervalo $(L - \epsilon, L + \epsilon)$), siempre es posible encontrar una franja vertical de ancho $2\delta$ centrada en $x_0$ (el intervalo $(x_0 - \delta, x_0 + \delta)$, excluyendo el propio $x_0$) tal que la gráfica de la función para todos los valores en ese intervalo vertical quede completamente contenida dentro de la franja horizontal.

### 5.3 Metodología Analítica para Encontrar $\delta(\epsilon)$ o Demostración $\epsilon-\delta$
Para demostrar formalmente un límite, debemos encontrar una regla explícita que nos dé un $\delta > 0$ adecuado para cualquier $\epsilon > 0$.
1. **Análisis Preliminar**: Partimos de la desigualdad deseada $|f(x) - L| < \epsilon$ e intentamos factorizarla para aislar el término de distancia $|x - x_0|$, obteniendo una expresión de la forma:
   $$|f(x) - L| = |g(x)| \cdot |x - x_0| < \epsilon$$
2. **Acotación de la Función Auxiliar $|g(x)|$**: Establecemos un entorno preliminar alrededor de $x_0$ asumiendo que $|x - x_0| < \delta_1$ (usualmente $\delta_1 = 1/2$ o $1$, o menor si hay asíntotas cercanas). Con esta restricción, encontramos un límite superior para la función auxiliar $|g(x)| \le M$ para alguna constante $M > 0$.
3. **Selección del $\delta$**: Si $|g(x)| \le M$, entonces $|g(x)| \cdot |x - x_0| < M \cdot |x - x_0|$. Para asegurar que esta cantidad sea menor que $\epsilon$, requerimos $|x - x_0| < \frac{\epsilon}{M}$ (llamado $\delta_2$).
4. **Conclusión**: Definimos $\delta = \min\{\delta_1, \delta_2\} = \min\{\delta_1, \frac{\epsilon}{M}\}$. De este modo, si $0 < |x - x_0| < \delta$, se cumplen simultáneamente ambas acotaciones y se garantiza la implicación.

### 5.4 Demostración Analítica Paso a Paso del Ejemplo del Texto
**Ejemplo**: Probar por definición que $\lim_{x \to -1} (3x^2 + x - 1) = 1$.

* **Identificación de elementos**:
  $$x_0 = -1, \quad f(x) = 3x^2 + x - 1, \quad L = 1$$
* **Paso 1: Desarrollar $|f(x) - L|$**:
  $$|f(x) - L| = |(3x^2 + x - 1) - 1| = |3x^2 + x - 2|$$
  Factorizando el polinomio cuadrático:
  $$3x^2 + x - 2 = (3x - 2)(x + 1)$$
  Por lo tanto:
  $$|f(x) - L| = |3x - 2| \cdot |x + 1| \quad (*)$$
  Aquí, el término $|x - x_0|$ es $|x - (-1)| = |x + 1|$.
* **Paso 2: Acotar el término $|3x - 2|$**:
  Asumimos una cota superior para la distancia a $x_0 = -1$. Elijamos un radio preliminar $\delta_1 = \frac{1}{2}$ de manera que:
  $$|x + 1| < \delta_1 = \frac{1}{2}$$
  Abriendo el valor absoluto:
  $$-\frac{1}{2} < x + 1 < \frac{1}{2} \implies -\frac{3}{2} < x < -\frac{1}{2}$$
  Multiplicamos por 3 en todas las partes de la desigualdad:
  $$-\frac{9}{2} < 3x < -\frac{3}{2}$$
  Restamos 2 para construir el término de interés $3x - 2$:
  $$-\frac{13}{2} < 3x - 2 < -\frac{7}{2}$$
  Por lo tanto, en magnitud el valor absoluto satisface:
  $$|3x - 2| < \frac{13}{2}$$
* **Paso 3: Definición del delta final**:
  Volviendo a $(*)$, si se cumplen ambas condiciones:
  $$|f(x) - L| = |3x - 2| \cdot |x + 1| < \frac{13}{2} |x + 1| < \epsilon$$
  Para garantizar la cota de $\epsilon$, requerimos $|x + 1| < \frac{2\epsilon}{13}$ (que llamamos $\delta_2$).
  
  El $\delta$ que verifica la demostración para cualquier $\epsilon > 0$ es:
  $$\delta = \min\left\{\frac{1}{2}, \ \frac{2\epsilon}{13}\right\}$$


```python
# Graficación de la Franja de Tolerancia Epsilon-Delta
def f_ejemplo(x):
    return 3*x**2 + x - 1

x_0 = -1.0
L = 1.0
epsilon = 0.8  # Fijamos una tolerancia epsilon

# Calculamos delta usando la fórmula analítica obtenida: delta = min(1/2, 2*epsilon/13)
delta_1 = 0.5
delta_2 = (2.0 * epsilon) / 13.0
delta = min(delta_1, delta_2)

print(f"Tolerancia Epsilon elegida = {epsilon}")
print(f"Cota delta_1 = {delta_1}")
print(f"Cota delta_2 (2*epsilon/13) = {delta_2:.4f}")
print(f"Radio Delta resultante = {delta:.4f}")

# Generación de puntos para la gráfica
x_plot = np.linspace(-1.6, -0.4, 400)
y_plot = f_ejemplo(x_plot)

plt.figure(figsize=(10, 7.5))

# Franja horizontal L +- epsilon
plt.fill_between(x_plot, L - epsilon, L + epsilon, color='coral', alpha=0.15, label='Franja horizontal $\\epsilon$ ($L \\pm \\epsilon$)')
plt.axhline(L, color='red', linestyle='--', alpha=0.8, label='$L = 1$')
plt.axhline(L - epsilon, color='red', linestyle=':', alpha=0.5)
plt.axhline(L + epsilon, color='red', linestyle=':', alpha=0.5)

# Franja vertical x_0 +- delta
plt.axvspan(x_0 - delta, x_0 + delta, color='skyblue', alpha=0.2, label='Entorno $\\delta$ ($x_0 \\pm \\delta$)')
plt.axvline(x_0, color='blue', linestyle='--', alpha=0.8, label='$x_0 = -1$')
plt.axvline(x_0 - delta, color='blue', linestyle=':', alpha=0.5)
plt.axvline(x_0 + delta, color='blue', linestyle=':', alpha=0.5)

# Gráfica de la función
plt.plot(x_plot, y_plot, 'k-', linewidth=2.5, label='$f(x) = 3x^2 + x - 1$')
plt.plot(x_0, L, 'ko', markersize=9)

# Dibujar el rectángulo de la demostración para verificar que no sale de la franja horizontal
rect_x = [x_0 - delta, x_0 + delta, x_0 + delta, x_0 - delta, x_0 - delta]
rect_y = [f_ejemplo(x_0 - delta), f_ejemplo(x_0 - delta), f_ejemplo(x_0 + delta), f_ejemplo(x_0 + delta), f_ejemplo(x_0 - delta)]
# Dibujamos la caja del límite
plt.plot(rect_x, rect_y, 'g', linewidth=2.0, linestyle='-.', label='Ventana de control $\\epsilon-\\delta$')

plt.title('Demostración Visual de la Definición del Límite $\\epsilon-\\delta$', fontweight='bold')
plt.xlabel('Variable independiente ($x$)')
plt.ylabel('Función ($f(x)$)')
plt.legend(loc='upper right')
plt.xlim(-1.55, -0.45)
plt.ylim(-0.2, 2.2)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 6. Continuidad y Límites Laterales

### 6.1 Límites Laterales
Cuando evaluamos la tendencia de una función $f(x)$ cerca de un punto $x_0$, podemos restringir la variable $x$ a aproximarse únicamente por un lado:
* **Límite por la Izquierda**: Se denota por $\lim_{x \to x_0^-} f(x) = L_1$. Analiza los valores de $f(x)$ para $x < x_0$.
* **Límite por la Derecha**: Se denota por $\lim_{x \to x_0^+} f(x) = L_2$. Analiza los valores de $f(x)$ para $x > x_0$.

> **Teorema de Existencia**: El límite global de una función en un punto $x_0$ existe y es igual a $L$ si y solo si ambos límites laterales existen y son iguales a $L$:
> $$\lim_{x \to x_0} f(x) = L \iff \lim_{x \to x_0^-} f(x) = \lim_{x \to x_0^+} f(x) = L$$

### 6.2 Continuidad de una Función en un Punto
Una función $f(x)$ es continua en un punto $a$ si su gráfica no presenta interrupciones o saltos en dicho punto. Formalmente, deben cumplirse tres condiciones indispensables:
1. **$f(a)$ existe** (es decir, $a$ pertenece al dominio de $f$).
2. **$\lim_{x \to a} f(x)$ existe** (es decir, los límites laterales son finitos e iguales).
3. **$\lim_{x \to a} f(x) = f(a)$** (el valor límite coincide con la evaluación de la función).

### 6.3 Clasificación de Discontinuidades
Si alguna de las tres condiciones falla, la función posee una **discontinuidad**:
* **Discontinuidad Evitable (Removible)**: El límite $\lim_{x \to a} f(x)$ existe y es finito, pero la función no está definida en $a$, o bien $f(a)$ difiere del valor del límite. Es posible "redefinir" la función en ese punto para hacerla continua.
* **Discontinuidad Esencial**: El límite global no existe.
  - **De Salto Finito**: Los límites laterales existen y son finitos, pero son diferentes (como en la función de Heaviside).
  - **De Salto Infinito**: Uno o ambos límites laterales tienden a $+\infty$ o $-\infty$.
  - **Oscilatoria**: La función oscila infinitamente al acercarse al punto (como en $\sin(\pi/x)$).

### 6.4 Límites Infinitos y Asíntotas Verticales
Consideremos la función racional:
$$f(x) = \frac{x}{x + 1}$$
Esta función no está definida en $x = -1$. Al analizar el comportamiento lateral de las imágenes cuando $x \to -1$:
* **Por la izquierda ($x < -1$)**: Los valores del denominador son negativos y muy cercanos a cero, haciendo que la fracción crezca en valor absoluto con signo positivo:
  $$\lim_{x \to -1^-} \frac{x}{x + 1} = +\infty$$
* **Por la derecha ($x > -1$)**: Los valores del denominador son positivos y muy cercanos a cero, haciendo que la fracción crezca en valor absoluto con signo negativo:
  $$\lim_{x \to -1^+} \frac{x}{x + 1} = -\infty$$
Dado que los límites laterales tienden a infinitos de signos opuestos, se dice que la recta $x = -1$ es una **asíntota vertical** de la función.


```python
# Función racional con asíntota
def f_rat(x):
    return x / (x + 1)

# 1. Aproximación numérica alrededor del punto de discontinuidad x = -1
x_as_izq = np.array([-1.1, -1.01, -1.001, -1.0001])
x_as_der = np.array([-0.9, -0.99, -0.999, -0.9999])

print("Aproximación por la Izquierda (x -> -1^-):")
print(f"{'x':<10} | {'f(x)':<12}")
print("-" * 25)
for xi, yi in zip(x_as_izq, f_rat(x_as_izq)):
    print(f"{xi:<10.4f} | {yi:<12.4f}")

print("\nAproximación por la Derecha (x -> -1^+):")
print(f"{'x':<10} | {'f(x)':<12}")
print("-" * 25)
for xi, yi in zip(x_as_der, f_rat(x_as_der)):
    print(f"{xi:<10.4f} | {yi:<12.4f}")

# 2. Graficación de la función racional y su asíntota vertical
x_plot_izq = np.linspace(-3.0, -1.05, 200)
x_plot_der = np.linspace(-0.95, 1.0, 200)

plt.figure(figsize=(10, 6))
plt.plot(x_plot_izq, f_rat(x_plot_izq), 'b-', linewidth=2.5, label='$f(x) = \\frac{x}{x+1}$')
plt.plot(x_plot_der, f_rat(x_plot_der), 'b-', linewidth=2.5)

# Asíntota vertical en x = -1
plt.axvline(-1.0, color='red', linestyle='--', linewidth=2.0, label='Asíntota Vertical $x = -1$')

plt.title('Discontinuidad Esencial por Salto Infinito', fontweight='bold')
plt.xlabel('Variable independiente ($x$)')
plt.ylabel('Función ($f(x)$)')
plt.legend()
plt.ylim(-10, 10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 7. Resumen de la Lección

1. **Concepto de Función**: Estructura de mapeo formal $f: A \to B$ que asigna una única imagen a cada elemento del dominio. Se clasifican según sus propiedades de mapeo en inyectivas, sobreyectivas y biyectivas.
2. **Función Lineal**: Representada por la ecuación de la recta $y = mx + b$. Se caracteriza por poseer una tasa de cambio (pendiente) constante $m$. Permite modelar leyes de oferta/demanda y el experimento cognitivo de Stenberg en medicina y psicología.
3. **Límite de una Función**: Describe el valor $L$ al que tienden las imágenes $f(x)$ a medida que $x$ se aproxima al punto $x_0$, sin requerir que la función esté definida en dicho punto.
4. **Inexistencia de Límites**: Los límites fallan al converger en presencia de saltos finitos de los límites laterales o al presentarse comportamientos oscilatorios de frecuencia infinita.
5. **Definición Formal ($\epsilon-\delta$)**: Formaliza de forma rigurosa la idea de aproximación a través de entornos y franjas de tolerancia, proporcionando metodologías algebraicas para demostrar límites concretos.
6. **Continuidad**: La continuidad exige que la función esté definida en el punto, que el límite global exista, y que ambos coincidan. Su incumplimiento da origen a discontinuidades clasificadas en evitables y esenciales.

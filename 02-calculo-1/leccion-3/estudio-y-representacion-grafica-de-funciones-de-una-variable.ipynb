## 1. Definición de Función y Formas de Representación

### 1.1 Definición de Función
Una **función** $f$ es una regla de correspondencia que asigna a cada elemento $x$ de un conjunto de partida $A$ exactamente un elemento, llamado $f(x)$, en un conjunto de llegada $B$. En el marco del Cálculo I, los conjuntos $A$ y $B$ son subconjuntos de los números reales $\mathbb{R}$ (funciones reales de variable real).

* **Variable Independiente**: El símbolo $x$, que representa un número arbitrario en el dominio de $f$.
* **Variable Dependiente**: El símbolo $y = f(x)$, que representa un número en el rango de $f$.
* **Dominio ($Dom(f)$)**: Conjunto de todas las entradas posibles $x$.
* **Rango o Imagen ($Im(f)$)**: Conjunto de todos los valores posibles de $y$ a medida que $x$ varía a través del dominio.

### 1.2 Las Cuatro Formas de Representación
Siguiendo a Stewart *et al.*, una función puede describirse de cuatro maneras distintas:
1. **Verbal**: Mediante una descripción en palabras (ej. *"el costo de enviar por correo una carta de primera clase en función de su peso en onzas"*).
2. **Algebraica**: Mediante una fórmula matemática explícita (ej. el área de un círculo en función de su radio: $A(r) = \pi r^2$).
3. **Visual**: Mediante una representación gráfica en el plano cartesiano (ej. la señal de un sismógrafo).
4. **Numérica**: Mediante una tabla de valores tabulados (ej. registros de población).

Frecuentemente, el análisis científico consiste en pasar de una representación a otra para comprender mejor el comportamiento físico modelado.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Caso de estudio: caída libre de una piedra d(t) = 16t^2 (t en segundos, d en pies)
t_valores = np.array([0, 1, 2, 3, 4])
d_valores = 16 * t_valores**2

print("Representación Numérica (Tabla de valores de caída libre):")
print(f"{'Tiempo t (s)':<12} | {'Distancia d(t) (pies)':<22}")
print("-" * 38)
for t, d in zip(t_valores, d_valores):
    print(f"{t:<12} | {d:<22}")
print("-" * 38)

# Representación Visual y Algebraica
t_plot = np.linspace(0, 4.2, 200)
d_plot = 16 * t_plot**2

plt.figure(figsize=(10, 6))
plt.plot(t_plot, d_plot, 'b-', linewidth=2.5, label='Fórmula algebraica: $d(t) = 16t^2$')
plt.scatter(t_valores, d_valores, color='red', s=60, zorder=5, label='Puntos medidos (tabla)')

plt.title('Representación Visual: Distancia de Caída Libre vs Tiempo', fontweight='bold')
plt.xlabel('Tiempo transcurrido ($t$, segundos)')
plt.ylabel('Distancia de caída ($d$, pies)')
plt.xlim(-0.2, 4.5)
plt.ylim(-15, 300)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 2. Clasificación de Funciones y Concepto de Asíntotas

### 2.1 Clasificación General de Funciones Escalares
Las funciones reales de variable real se organizan en familias según sus operaciones matemáticas:
* **Funciones Algebraicas**: La variable independiente $x$ está afectada únicamente por operaciones algebraicas básicas (suma, resta, producto, cociente, potencia y radicación).
  - *Racionales Enteras*: Los polinomios.
  - *Racionales Fraccionarias*: Cocientes de polinomios.
  - *Irracionales*: Contienen variables bajo signos de radicación.
* **Funciones Trascendentes**: No pueden expresarse mediante combinaciones finitas de operaciones algebraicas (exponenciales, logarítmicas, trigonométricas, hiperbólicas e inversas).

### 2.2 Definición Formal de Asíntotas
Las asíntotas son rectas a las que se aproxima la curva de la función de manera indefinida a medida que nos desplazamos por su dominio:

1. **Asíntota Vertical ($x = a$)**:
   La recta vertical $x = a$ es una asíntota vertical de la curva $y = f(x)$ si al menos una de las siguientes afirmaciones de límites laterales es cierta:
   $$\lim_{x \to a^-} f(x) = \pm\infty \quad \text{o} \quad \lim_{x \to a^+} f(x) = \pm\infty$$
   
2. **Asíntota Horizontal ($y = b$)**:
   La recta horizontal $y = b$ es una asíntota horizontal de la curva $y = f(x)$ si se cumple el límite en el infinito:
   $$\lim_{x \to -\infty} f(x) = b \quad \text{o} \quad \lim_{x \to +\infty} f(x) = b$$


```python
# Gráfica de la función f(x) = 1/x y sus asíntotas
x_izq = np.linspace(-4.0, -0.05, 200)
x_der = np.linspace(0.05, 4.0, 200)

plt.figure(figsize=(10, 6))
plt.plot(x_izq, 1/x_izq, 'b-', linewidth=2.5, label='$f(x) = 1/x$')
plt.plot(x_der, 1/x_der, 'b-', linewidth=2.5)

# Asíntota Vertical x = 0 (Eje Y)
plt.axvline(0, color='red', linestyle='--', linewidth=1.8, label='Asíntota Vertical $x = 0$')
# Asíntota Horizontal y = 0 (Eje X)
plt.axhline(0, color='green', linestyle='-.', linewidth=1.8, label='Asíntota Horizontal $y = 0$')

plt.title('Representación de Asíntotas para la Función Recíproca $f(x) = 1/x$', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-6, 6)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 3. Estudio de Funciones Polinómicas

### 3.1 Definición Matemática de Polinomio
Una **función polinomial** de grado $n$ es una función de la forma:
$$P(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0$$
donde $n$ es un entero no negativo y el coeficiente principal $a_n \neq 0$. Los números reales $a_k$ se llaman coeficientes del polinomio.

### 3.2 Trazado y Características de las Funciones Polinómicas
Las gráficas polinómicas presentan propiedades de suavidad y continuidad:
* **Suavidad**: Son curvas continuas y suaves (sin picos, esquinas ni discontinuidades en su dominio $\mathbb{R}$).
* **Monomios $x^n$**: Si $n$ es par, la gráfica es simétrica al eje $Y$ (forma similar a una parábola). Si $n$ es impar, es simétrica al origen (forma similar a $y = x^3$). Conforme aumenta $n$, la curva se aplana en el intervalo $(-1, 1)$ y se empina rápidamente fuera de él.
* **Extremos Locales**: Por el cálculo diferencial, un polinomio de grado $n$ puede tener **a lo sumo $n-1$ extremos locales** (máximos o mínimos).

### 3.3 Reglas Generales para Graficar Polinomios
1. **Ceros (Raíces)**: Factorizar el polinomio para encontrar sus ceros reales (puntos de intersección con el eje $x$, donde $P(x) = 0$).
2. **Puntos de Prueba**: Evaluar puntos intermedios entre las raíces para determinar si la función yace por encima o por debajo del eje $x$ (signo del polinomio).
3. **Comportamiento Extremo**: Determinar el límite cuando $x \to \pm\infty$, el cual está gobernado enteramente por su término principal $a_n x^n$.
4. **Trazado**: Unir los puntos característicos mediante una curva lisa y continua.


```python
# Trazado de un Polinomio de Grado 5: P(x) = 3x^5 + 6x^4 - 2x^3 + x^2 + 7x - 6
def P_pol(x):
    return 3*x**5 + 6*x**4 - 2*x**3 + x**2 + 7*x - 6

x_pol = np.linspace(-2.2, 1.2, 400)
y_pol = P_pol(x_pol)

# Encontrar numéricamente las raíces reales de P(x) para marcarlas
coefs_pol = [3, 6, -2, 1, 7, -6]
raices_reales = np.roots(coefs_pol)
raices_reales = sorted([r.real for r in raices_reales if np.isreal(r)])

print("Análisis del Polinomio:")
print(f"  Grado: 5, Coeficiente Principal: 3, Término Constante: -6")
print(f"  Raíces reales halladas: {raices_reales}")
print(f"  Máximo número de extremos locales posibles (n-1): 4")
print("-" * 60)

plt.figure(figsize=(10, 6.5))
plt.plot(x_pol, y_pol, 'b-', linewidth=2.5, label='$P(x) = 3x^5 + 6x^4 - 2x^3 + x^2 + 7x - 6$')

# Marcar intersecciones con el eje X (raíces)
plt.scatter(raices_reales, [0]*len(raices_reales), color='red', s=70, zorder=5, label='Raíces reales')
# Marcar intersección con el eje Y (P(0) = -6)
plt.scatter(0, -6.0, color='darkgreen', s=70, zorder=5, label='Intersección Y $(0, -6)$')

plt.title('Representación Gráfica de un Polinomio de Grado 5', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$P(x)$')
plt.xlim(-2.3, 1.3)
plt.ylim(-20, 20)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. Estudio de Funciones Racionales

### 4.1 Definición Matemática
Una **función racional** tiene la forma:
$$r(x) = \frac{P(x)}{Q(x)}$$
donde $P(x)$ y $Q(x)$ son polinomios sin factores comunes de grado $n$ y $m$ respectivamente.
El dominio consiste en todos los números reales excepto los valores donde $Q(x) = 0$.

### 4.2 Determinación Analítica de Asíntotas en Funciones Racionales
El comportamiento en los extremos y las singularidades de la función racional siguen reglas algebraicas precisas según sus grados:

1. **Asíntotas Verticales**: Son las rectas verticales $x = a$, donde $a$ es un cero del denominador ($Q(a) = 0$).
2. **Asíntota Horizontal**:
   - **Caso $n < m$**: La recta $y = 0$ (el eje $x$) es la asíntota horizontal.
   - **Caso $n = m$**: La recta $y = \frac{a_n}{b_m}$ es la asíntota horizontal, siendo $a_n$ y $b_m$ los coeficientes principales de $P$ y $Q$.
   - **Caso $n > m$**: La función no posee asíntota horizontal (puede tener una asíntota oblicua si $n = m+1$).


```python
# Gráfica de la función racional r(x) = 2x^2 / (x^2 - 9)
# Grado del numerador (n) = 2, Grado del denominador (m) = 2. Asíntota horizontal: y = 2/1 = 2.
# Ceros del denominador: x = 3, -3 (asíntotas verticales).
def r_func(x):
    return (2 * x**2) / (x**2 - 9)

x_t1 = np.linspace(-8.0, -3.05, 200)
x_t2 = np.linspace(-2.95, 2.95, 200)
x_t3 = np.linspace(3.05, 8.0, 200)

plt.figure(figsize=(11, 7.5))
plt.plot(x_t1, r_func(x_t1), 'b-', linewidth=2.5, label='$r(x) = \\frac{2x^2}{x^2-9}$')
plt.plot(x_t2, r_func(x_t2), 'b-', linewidth=2.5)
plt.plot(x_t3, r_func(x_t3), 'b-', linewidth=2.5)

# Dibujar asíntotas
plt.axvline(-3.0, color='red', linestyle='--', linewidth=1.8, label='Asíntotas Verticales $x = \\pm 3$')
plt.axvline(3.0, color='red', linestyle='--')
plt.axhline(2.0, color='green', linestyle='-.', linewidth=1.8, label='Asíntota Horizontal $y = 2$')
plt.scatter(0, 0, color='darkgreen', s=60, zorder=5, label='Origen (0, 0)')

plt.title('Estudio de la Función Racional $r(x) = \\frac{2x^2}{x^2-9}$ y sus Asíntotas', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-8.0, 8.0)
plt.ylim(-8.0, 8.0)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 5. Estudio de Funciones Irracionales

### 5.1 Definición Matemática
Una **función irracional** contiene la variable independiente $x$ bajo un radical de la forma:
$$f(x) = \sqrt[n]{g(x)}$$

### 5.2 Determinación del Dominio según el Índice de Raíz
El comportamiento del dominio depende enteramente de la paridad de la raíz:
* **Índice Par ($n$ es par)**: Para que las imágenes sean números reales, el radicando no puede ser negativo, por lo que requerimos:
  $$g(x) \ge 0$$
* **Índice Impar ($n$ es impar)**: El radical está definido para cualquier número real. El dominio de la función es idéntico al dominio de la función interna $g(x)$. Si $g(x)$ es racional, su dominio es $\mathbb{R}$:
  $$Dom(f) = (-\infty, +\infty)$$

### 5.3 Análisis de los 4 Ejemplos Clásicos de la Literatura
Graficamos y contrastamos las cuatro funciones irracionales provistas en el material teórico:

a. **Raíz par simple**: $f(x) = \sqrt{2x + 5}$. Radicando $2x+5 \ge 0 \implies x \ge -\frac{5}{2}$. Dominio $Dom(f) = [-\frac{5}{2}, +\infty)$.
b. **Raíz par compuesta**: $f(x) = \sqrt[6]{x^2 - 4}$. Radicando $x^2 - 4 \ge 0 \implies x \le -2$ o $x \ge 2$. Dominio $Dom(f) = (-\infty, -2] \cup [2, +\infty)$.
c. **Raíz impar compuesta (cúbica)**: $f(x) = \sqrt[3]{x^2 - 1}$. Índice impar. Dominio $Dom(f) = (-\infty, +\infty)$. Presenta un punto singular en $x = \pm 1$ con pendiente vertical.
d. **Raíz impar compuesta (quinta)**: $f(x) = \sqrt[5]{3x + 2}$. Índice impar. Dominio $Dom(f) = (-\infty, +\infty)$.


```python
# Creamos un panel 2x2 para representar los 4 ejemplos de funciones irracionales
fig, axs = plt.subplots(2, 2, figsize=(14, 11))

# Subplot a: f(x) = sqrt(2x+5) (Dominio: x >= -2.5)
x_a = np.linspace(-2.5, 8.0, 300)
y_a = np.sqrt(2 * x_a + 5)
axs[0, 0].plot(x_a, y_a, 'b-', linewidth=2.0, label='$f(x) = \\sqrt{2x+5}$')
axs[0, 0].scatter(-2.5, 0, color='red', s=50, zorder=5, label='Borde del Dominio')
axs[0, 0].set_title('a) Raíz par simple: $Dom(f) = [-2.5, +\\infty)$', fontweight='bold')
axs[0, 0].set_xlim(-3.5, 8.0)
axs[0, 0].set_ylim(-1.0, 5.0)
axs[0, 0].legend(loc='upper left')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Subplot b: f(x) = (x^2-4)^(1/6) (Dominio: |x| >= 2)
x_b_izq = np.linspace(-8.0, -2.0, 300)
x_b_der = np.linspace(2.0, 8.0, 300)
y_b_izq = (x_b_izq**2 - 4)**(1/6)
y_b_der = (x_b_der**2 - 4)**(1/6)
axs[0, 1].plot(x_b_izq, y_b_izq, 'g-', linewidth=2.0, label='$f(x) = \\sqrt[6]{x^2-4}$')
axs[0, 1].plot(x_b_der, y_b_der, 'g-', linewidth=2.0)
axs[0, 1].scatter([-2, 2], [0, 0], color='red', s=50, zorder=5, label='Bordes')
axs[0, 1].set_title('b) Raíz par compuesta: $Dom(f) = (-\\infty, -2] \\cup [2, +\\infty)$', fontweight='bold')
axs[0, 1].set_xlim(-8.5, 8.5)
axs[0, 1].set_ylim(-0.5, 3.0)
axs[0, 1].legend(loc='upper center')
axs[0, 1].grid(True, linestyle='--', alpha=0.7)

# Subplot c: f(x) = (x^2-1)^(1/3) (Dominio: R)
x_c = np.linspace(-6.0, 6.0, 500)
y_c = np.sign(x_c**2 - 1) * np.abs(x_c**2 - 1)**(1/3)
axs[1, 0].plot(x_c, y_c, 'crimson', linewidth=2.0, label='$f(x) = \\sqrt[3]{x^2-1}$')
axs[1, 0].scatter([-1, 1], [0, 0], color='red', s=50, zorder=5, label='Puntos singulares')
axs[1, 0].set_title('c) Raíz impar: $Dom(f) = (-\\infty, +\\infty)$', fontweight='bold')
axs[1, 0].set_xlim(-6.5, 6.5)
axs[1, 0].set_ylim(-2.5, 4.0)
axs[1, 0].legend(loc='upper center')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Subplot d: f(x) = (3x+2)^(1/5) (Dominio: R)
x_d = np.linspace(-6.0, 6.0, 500)
y_d = np.sign(3 * x_d + 2) * np.abs(3 * x_d + 2)**(1/5)
axs[1, 1].plot(x_d, y_d, 'darkcyan', linewidth=2.0, label='$f(x) = \\sqrt[5]{3x+2}$')
axs[1, 1].scatter(-2/3, 0, color='red', s=50, zorder=5, label='Punto singular')
axs[1, 1].set_title('d) Raíz impar simple: $Dom(f) = (-\\infty, +\\infty)$', fontweight='bold')
axs[1, 1].set_xlim(-6.5, 6.5)
axs[1, 1].set_ylim(-2.5, 2.5)
axs[1, 1].legend(loc='upper left')
axs[1, 1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 6. Estudio de Funciones Exponenciales y Logarítmicas

### 6.1 Función Exponencial
La **función exponencial** con base $a$ se define para todos los números reales $x$ mediante:
$$f(x) = a^x \quad (a > 0, \ a \neq 1)$$

* **Dominio y Rango**: $Dom(f) = \mathbb{R}$, $Im(f) = (0, +\infty)$ (las imágenes son estrictamente positivas).
* **Asíntota Horizontal**: La recta $y = 0$ (el eje $x$) es una asíntota horizontal de la función exponencial.
  - Si $a > 1$, la función crece rápidamente: $\lim_{x \to -\infty} a^x = 0$.
  - Si $0 < a < 1$, la función decrece rápidamente: $\lim_{x \to +\infty} a^x = 0$.
* **Uno a Uno**: Puesto que es una función estrictamente monótona, pasa el test de la recta horizontal y, por tanto, admite una función inversa única.

### 6.2 Función Logarítmica
La **función logarítmica** con base $a$ es la función inversa de la exponencial:
$$\log_a x = y \iff a^y = x$$

* **Dominio y Rango**: El dominio del logaritmo es el rango de la exponencial, $Dom(\log_a) = (0, +\infty)$. El rango es $\mathbb{R}$.
* **Asíntota Vertical**: La recta $x = 0$ (el eje $y$) es una asíntota vertical, debido a que:
  $$\lim_{x \to 0^+} \log_a x = -\infty \quad (\text{para } a > 1)$$
* **Intersección**: La gráfica de toda función logarítmica interseca el eje $x$ en el punto $(1, 0)$, puesto que $\log_a 1 = 0$ para cualquier base $a$.

### 6.3 Simetría Reflexiva de las Funciones Inversas
Geométricamente, la gráfica de una función inversa $f^{-1}$ se obtiene al reflejar la gráfica de $f$ sobre la recta identidad $y = x$. Así, las curvas de $y = a^x$ e $y = \log_a x$ son perfectamente simétricas respecto a dicha bisectriz.


```python
# Graficación de funciones inversas: e^x y ln(x) mostrando la simetría sobre y = x
x_exp = np.linspace(-3.5, 2.5, 200)
y_exp = np.exp(x_exp)

x_log = np.linspace(0.03, 8.5, 300)
y_log = np.log(x_log)

x_id = np.linspace(-3.5, 8.5, 100)
y_id = x_id

plt.figure(figsize=(9.5, 9.5))
plt.plot(x_exp, y_exp, 'crimson', linewidth=2.5, label='Función Exponencial $y = e^x$')
plt.plot(x_log, y_log, 'darkcyan', linewidth=2.5, label='Función Logarítmica $y = \\ln(x)$')
plt.plot(x_id, y_id, 'gray', linestyle='--', linewidth=1.5, label='Bisectriz $y = x$')

# Marcar intersecciones y puntos simétricos correspondientes
plt.scatter([0, 1], [1, 0], color='black', s=55, zorder=5)
plt.text(0.2, 1.2, '$(0, 1)$', fontweight='bold')
plt.text(1.2, 0.2, '$(1, 0)$', fontweight='bold')

# Asíntotas
plt.axhline(0, color='crimson', linestyle=':', alpha=0.5, label='A. Horizontal $y = 0$ ($e^x$)')
plt.axvline(0, color='darkcyan', linestyle=':', alpha=0.5, label='A. Vertical $x = 0$ ($\\ln x$)')

plt.title('Simetría de Funciones Inversas: Exponencial y Logarítmica Natural', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-3.0, 8.0)
plt.ylim(-3.0, 8.0)
plt.gca().set_aspect('equal', adjustable='box') # Proporción 1:1 real
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 7. Estudio de Funciones Trigonométricas

### 7.1 Definición sobre el Círculo Unitario
Las funciones trigonométricas se definen geométricamente sobre una circunferencia de radio unitario $x^2 + y^2 = 1$ en el plano cartesiano:
Si $t$ es la longitud de un arco recorrido en sentido antihorario a partir del punto $(1, 0)$ sobre el círculo, las coordenadas del punto terminal $P(x, y)$ determinan las razones trigonométricas:
$$\cos t = x, \quad \sin t = y, \quad \tan t = \frac{y}{x} \ \ (x \neq 0)$$
$$\sec t = \frac{1}{x} \ \ (x \neq 0), \quad \csc t = \frac{1}{y} \ \ (y \neq 0), \quad \cot t = \frac{x}{y} \ \ (y \neq 0)$$

### 7.2 Comportamiento por Cuadrantes (Regla Nemotécnica)
El signo de cada función depende de las coordenadas $(x, y)$ del cuadrante del punto terminal $P$:

| Cuadrante | Rango del Ángulo | Coordenadas | Funciones Positivas | Funciones Negativas |
| :---: | :---: | :---: | :---: | :--- |
| **I** | $0$ a $\pi/2$ | $(+, +)$ | **Todas** | Ninguna |
| **II** | $\pi/2$ a $\pi$ | $(-, +)$ | **Seno**, Cosecante | Coseno, Secante, Tangente, Cotangente |
| **III** | $\pi$ a $3\pi/2$ | $(-, -)$ | **Tangente**, Cotangente | Seno, Cosecante, Coseno, Secante |
| **IV** | $3\pi/2$ a $2\pi$ | $(+, -)$ | **Coseno**, Secante | Seno, Cosecante, Tangente, Cotangente |

Nemotecnia clásica: **"Todas, Seno, Tangente, Coseno"**.


```python
# Visualización del Círculo Unitario y la onda senoidal correspondiente
theta = np.linspace(0, 2*np.pi, 300)

fig, (ax_circ, ax_sine) = plt.subplots(1, 2, figsize=(14, 6))

# 1. Círculo Unitario
ax_circ.plot(np.cos(theta), np.sin(theta), 'k-', linewidth=2.0, label='Circunferencia $x^2 + y^2 = 1$')
ax_circ.axhline(0, color='gray', linewidth=0.8)
ax_circ.axvline(0, color='gray', linewidth=0.8)

# Dibujar ángulo de ejemplo en t = pi/3 (60 grados)
t_ej = np.pi / 3
x_ej = np.cos(t_ej)
y_ej = np.sin(t_ej)
ax_circ.plot([0, x_ej], [0, y_ej], 'r-o', linewidth=2.0)
ax_circ.plot([x_ej, x_ej], [0, y_ej], 'g--')
ax_circ.plot([0, x_ej], [y_ej, y_ej], 'b--')
ax_circ.text(0.1, 0.08, '$\theta$', color='red', fontsize=13, fontweight='bold')
ax_circ.text(x_ej/2, -0.15, '$x = \\cos\\theta$', color='blue', ha='center')
ax_circ.text(x_ej + 0.05, y_ej/2, '$y = \\sin\\theta$', color='green')

ax_circ.set_title('Círculo Unitario y Proyección Trigonométrica', fontweight='bold')
ax_circ.set_xlim(-1.25, 1.25)
ax_circ.set_ylim(-1.25, 1.25)
ax_circ.set_aspect('equal', adjustable='box')
ax_circ.legend(loc='lower left')

# 2. Onda Seno correspondiente
ax_sine.plot(theta, np.sin(theta), 'g-', linewidth=2.5, label='$y = \\sin(\\theta)$')
ax_sine.plot([t_ej, t_ej], [0, y_ej], 'r--')
ax_sine.plot(t_ej, y_ej, 'ro', markersize=8)
ax_sine.axhline(0, color='gray', linewidth=0.8)

ax_sine.set_title('Trazado de la Función Seno', fontweight='bold')
ax_sine.set_xlabel('Ángulo $\\theta$ (radianes)')
ax_sine.set_ylabel('Amplitud')
ax_sine.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
ax_sine.set_xticklabels(['0', '$\\pi/2$', '$\\pi$', '$3\\pi/2$', '$2\\pi$'])
ax_sine.set_xlim(0, 2*np.pi)
ax_sine.set_ylim(-1.2, 1.2)
ax_sine.legend()
ax_sine.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 8. Combinación y Composición de Funciones

### 8.1 Combinación Aritmética de Funciones (Álgebra de Funciones)
Dadas dos funciones $f$ y $g$ con dominios $A$ y $B$, se definen las nuevas funciones:
* **Suma**: $(f + g)(x) = f(x) + g(x)$
* **Diferencia**: $(f - g)(x) = f(x) - g(x)$
* **Producto**: $(fg)(x) = f(x) \cdot g(x)$
El dominio común de estas tres operaciones es la intersección de los dominios individuales:
$$Dom = A \cap B$$
* **Cociente**: $\left(\frac{f}{g}\right)(x) = \frac{f(x)}{g(x)}$
El dominio consiste en los puntos comunes donde el denominador no se anule:
$$Dom = \{x \in A \cap B \mid g(x) \neq 0\}$$

### 8.2 Composición de Funciones (Función Compuesta)
La **composición** de la función $f$ con la función $g$ (denotada por $f \circ g$) es una operación secuencial donde la salida de la primera función sirve como entrada de la segunda:
$$(f \circ g)(x) = f(g(x))$$

El dominio de la función compuesta $f \circ g$ es el conjunto de todos los valores de $x$ en el dominio de $g$ tales que su imagen $g(x)$ pertenece al dominio de $f$:
$$Dom(f \circ g) = \{x \in Dom(g) \mid g(x) \in Dom(f)\}$$

**Ejemplo de Composición (Stewart *et al.*)**:
Si $f(x) = \sqrt{x}$ (cuyo dominio es $x \ge 0$) y $g(x) = x^2 + 1$ (cuyo dominio es $\mathbb{R}$):
$$h(x) = (f \circ g)(x) = f(g(x)) = f(x^2 + 1) = \sqrt{x^2 + 1}$$
Dado que para todo número real $x^2 + 1 \ge 1 > 0$, la función interna entrega siempre valores válidos para la raíz cuadrada externa. El dominio de la compuesta es toda la recta real:
$$Dom(h) = (-\infty, +\infty)$$


```python
# Visualización del flujo de composición de funciones: x -> g(x) = x^2 + 1 -> f(g(x)) = sqrt(x^2 + 1)
x_vals = np.linspace(-3.0, 3.0, 200)
y_g = x_vals**2 + 1
y_comp = np.sqrt(y_g)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 1. Función interna g(x) = x^2 + 1
ax1.plot(x_vals, y_g, 'g-', linewidth=2.5, label='$g(x) = x^2 + 1$')
ax1.set_title('Función Interna (Fase 1: Entrada $x \\to g(x)$)', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$g(x)$')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Función compuesta f(g(x)) = sqrt(x^2 + 1)
ax2.plot(x_vals, y_comp, 'b-', linewidth=2.5, label='$f(g(x)) = \\sqrt{x^2+1}$')
ax2.set_title('Función Compuesta Final (Fase 2: Salida $g(x) \\to f(g(x))$)', fontweight='bold')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_ylim(0.5, 3.5)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 9. Resumen de la Lección

1. **Representación de Funciones**: Las funciones son mapeos lógicos representables mediante cuatro descripciones complementarias: verbales, algebraicas, visuales y numéricas.
2. **Asíntotas**: Son rectas que describen el comportamiento de funciones al infinito (asíntotas horizontales $y = b$) o en las cercanías de discontinuidades del dominio (asíntotas verticales $x = a$).
3. **Polinomios**: Son funciones continuas y suaves caracterizadas por su grado y coeficientes. La presencia de raíces y su comportamiento extremo define su trazado geométrico básico.
4. **Funciones Racionales**: Se definen como cocientes de polinomios, mostrando un comportamiento asintótico complejo y vertical gobernado por los ceros de su denominador.
5. **Funciones Irracionales**: Requieren análisis de paridad en el índice de la raíz para determinar restricciones de dominio, admitiendo imágenes en todo $\mathbb{R}$ para índices impares.
6. **Trascendentes Inversas**: Las curvas exponenciales y logarítmicas ilustran de manera perfecta la simetría reflexiva respecto de la recta identidad $y = x$.
7. **Funciones Circulares**: Las razones trigonométricas proyectan la geometría circular sobre formas de onda continuas de carácter periódico.
8. **Composición de Funciones**: La composición $(f \circ g)(x)$ representa una cadena de transformaciones secuenciales cuyo dominio se ve condicionado por el rango de la función interna y el dominio de la función externa.

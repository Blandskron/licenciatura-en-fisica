## 1. Concepto de Primitiva de una Función e Integral Indefinida

### 1.1 Función Primitiva o Antiderivada
Se llama **función primitiva** (o antiderivada) de una función $f$ a otra función $P$ tal que para todo punto $x$ perteneciente al dominio de $f$, se cumple la igualdad de su derivada:
$$P'(x) = f(x) \quad \forall x \in Dom_f$$

Por ejemplo, la función $P(x) = \sin(x)$ es una primitiva de $f(x) = \cos(x)$, puesto que $(\sin(x))' = \cos(x)$. De igual manera, $P(x) = \arctan(x)$ es primitiva de $f(x) = \frac{1}{1+x^2}$, y $P(x) = \frac{x^3}{3}$ es primitiva de $f(x) = x^2$.

### 1.2 Teorema de Unicidad Salvo Constante
> **Teorema**: Dos funciones $P(x)$ y $H(x)$ son primitivas de una misma función $f(x)$ en el intervalo $(a, b)$ si y solo si difieren en una constante real.
>
> En símbolos:
> $$P(x) \text{ y } H(x) \text{ primitivas de } f \text{ en } (a, b) \iff P(x) - H(x) = C, \quad \forall x \in (a, b) \text{ y } C \in \mathbb{R}$$

* **Demostración**:
  1. *Dirección Directa ($\implies$)*: Si $P$ y $H$ son primitivas, por definición $P'(x) = f(x)$ y $H'(x) = f(x)$. Definiendo la función diferencia $g(x) = P(x) - H(x)$, su derivada es:
     $$g'(x) = P'(x) - H'(x) = f(x) - f(x) = 0$$
     Dado que la derivada es cero en todo el intervalo, por el Teorema del Valor Medio, la función diferencia es constante: $g(x) = C \implies P(x) - H(x) = C$.
  2. *Dirección Inversa ($\impliedby$)*: Si $P(x) = H(x) + C$, derivando obtenemos $P'(x) = (H(x) + C)' = H'(x) + 0 = f(x)$, por lo que $P$ también es primitiva de $f$.

### 1.3 Integral Indefinida
Al conjunto formado por todas las infinitas primitivas $P(x) + C$ de una función $f(x)$ se le llama **integral indefinida** de $f$, y se representa mediante:
$$\int f(x) dx = P(x) + C \iff [P(x) + C]' = f(x)$$

Donde los elementos se nombran:
* $\int$: símbolo integral.
* $f(x)$: función integrando o función a integrar.
* $dx$: diferencial de $x$, que indica la variable de integración.
* $C$: constante de integración (un número real arbitrario).


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Graficación de la familia de curvas de la integral indefinida de cos(x) -> sin(x) + C
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 300)
y_base = np.sin(x_vals)
C_list = [-3, -2, -1, 0, 1, 2, 3]

plt.figure(figsize=(11, 7.5))

# Graficar f(x) = cos(x) como referencia del integrando
plt.plot(x_vals, np.cos(x_vals), 'k--', linewidth=1.5, label='Integrando: $f(x) = \\cos(x)$', alpha=0.5)

# Graficar la familia de primitivas sin(x) + C
colores = plt.cm.viridis(np.linspace(0.1, 0.9, len(C_list)))
for C, col in zip(C_list, colores):
    label_text = f'Primitiva $P(x) = \\sin(x) + {C}$' if C >= 0 else f'Primitiva $P(x) = \\sin(x) - {abs(C)}$'
    plt.plot(x_vals, y_base + C, color=col, linewidth=2.0, label=label_text)

plt.title('Familia de Curvas de la Integral Indefinida: $\\int \\cos(x) dx = \\sin(x) + C$', fontweight='bold')
plt.xlabel('Variable independiente ($x$)')
plt.ylabel('Primitivas ($P(x)$)')
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(-4.5, 4.5)
plt.legend(loc='upper right', ncol=2)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 2. Propiedades de la Integral Indefinida y Diferenciales

### 2.1 Propiedades Operacionales de la Integral
La integral indefinida y el operador derivada $D$ interactúan bajo las siguientes leyes fundamentales:

1. **P1 (Derivada de la integral)**: Si $f$ es integrable, el operador derivada deshace la integral recuperando el integrando:
   $$D \left[ \int f(x) dx \right] = \frac{d}{dx} \left[ P(x) + C \right] = P'(x) = f(x)$$
   *Ejemplo*: $D \left[ \int \tan(5x) dx \right] = \tan(5x)$.
2. **P2 (Integral de la derivada)**: Si $f$ es derivable, la integral de la derivada recupera la función más una constante:
   $$\int D[f(x)] dx = \int f'(x) dx = f(x) + C$$
   *Ejemplo*: $\int D(\tan(5x)) dx = \tan(5x) + C$.
3. **P3 (Linealidad)**: La integral indefinida es un operador lineal sobre sumas y factores escalares:
   $$\dots \int \left( \alpha f(x) \pm \beta g(x) \right) dx = \alpha \int f(x) dx \pm \beta \int g(x) dx$$

### 2.2 Diferencial de una Función e Interpretación Geométrica
Sea $y = f(x)$ una función derivable en el intervalo $(a, b)$.
* **Definición del Diferencial ($dy$ o $df(x)$)**: Se define como el producto de la derivada de la función por el incremento de la variable independiente $\Delta x$:
   $$dy = df(x) = f'(x) \cdot \Delta x$$
   Si aplicamos la definición a la función identidad $f(x) = x$, su derivada es $f'(x) = 1 \implies dx = 1 \cdot \Delta x = \Delta x$.
   Por lo tanto, podemos reescribir de forma general:
   $$dy = f'(x) \cdot dx$$
   Esto nos permite formalizar la derivada como una razón de diferenciales: $f'(x) = \frac{dy}{dx}$.

* **Aproximación Lineal**: Geométricamente, el diferencial $dy$ representa el incremento de la ordenada de la recta tangente correspondiente a un incremento $dx$. Permite aproximar el valor de la función en puntos cercanos a $x_0$:
  $$f(x_0 + \Delta x) \approx f(x_0) + df(x_0) = f(x_0) + f'(x_0) dx$$


```python
# Simulación de Aproximación Lineal usando Diferenciales
# f(x) = ln(x) en torno a x_0 = 1.0 (donde ln(1) = 0 y f'(1) = 1/1 = 1)
def f_diff(x):
    return np.log(x)

x_0 = 1.0
y_0 = f_diff(x_0)
df_0 = 1.0 / x_0  # f'(x_0) = 1/x_0

dx_list = np.array([0.5, 0.2, 0.1, 0.05, 0.01])

print("Aproximación lineal por diferencial de ln(x_0 + dx):")
print(f"{'dx':<10} | {'Valor Exacto':<15} | {'Valor Aprox.':<15} | {'Error Absoluto':<15}")
print("-" * 63)
for dx in dx_list:
    val_real = f_diff(x_0 + dx)
    # Aproximación: f(x_0) + f'(x_0)*dx
    val_aprox = y_0 + df_0 * dx
    err = abs(val_real - val_aprox)
    print(f"{dx:<10.2f} | {val_real:<15.8f} | {val_aprox:<15.8f} | {err:<15.8e}")
```

## 3. Métodos de Integración: Por Partes y Cambio de Variable

### 3.1 Integración por Cambio de Variable (Sustitución)
Este método se basa en la regla de la cadena para derivadas. Si hacemos el cambio de variable $u = g(x)$, su diferencial es $du = g'(x) dx$:
$$\int f(g(x)) g'(x) dx = \int f(u) du = P(u) + C = P(g(x)) + C$$

### 3.2 Integración por Partes
Este método se deriva directamente de la regla para la derivada del producto de dos funciones:
$$\frac{d}{dx} [f(x) \cdot g(x)] = f'(x)g(x) + f(x)g'(x)$$
En términos de diferenciales, integrando en ambos lados:
$$\int d[f(x) \cdot g(x)] = \int f'(x)g(x) dx + \int f(x)g'(x) dx$$
$$f(x) \cdot g(x) = \int g(x) df(x) + \int f(x) dg(x)$$

Haciendo $u = f(x) \implies du = df(x)$, y $v = g(x) \implies dv = dg(x)$, despejamos una de las integrales obteniendo la fórmula clásica:
$$\int u dv = u v - \int v du$$

La clave del método radica en seleccionar adecuadamente los términos $u$ y $dv$ (usualmente bajo la nemotecnia *ILATE*: Inversas, Logarítmicas, Algebricas, Trigonométricas, Exponenciales), de modo que la integral resultante $\int v du$ sea más sencilla de evaluar que la original.


```python
import sympy as sp

x = sp.Symbol('x')

# 1. Integración por partes: int x * e^x dx (u = x, dv = e^x dx -> du = dx, v = e^x)
# Resultado: x*e^x - int e^x dx = (x-1)*e^x + C
f_partes = x * sp.exp(x)
int_partes = sp.integrate(f_partes, x)

print("1. Método por Partes (int x * e^x dx):")
print(f"  Integrando: f(x) = {f_partes}")
print(f"  Resultado analítico: {int_partes} + C")
print("-" * 60)

# 2. Método por cambio de variable: int ln(x)/x dx (u = ln(x) -> du = 1/x dx)
# Resultado: int u du = u^2/2 = (ln(x))^2 / 2 + C
f_sust = sp.log(x) / x
int_sust = sp.integrate(f_sust, x)

print("2. Método por Cambio de Variable (int ln(x)/x dx):")
print(f"  Integrando: f(x) = {f_sust}")
print(f"  Resultado analítico: {int_sust} + C")
```

## 4. Integración de Funciones Racionales

Consideramos la integral de un cociente de polinomios:
$$\int \frac{P(x)}{Q(x)} dx$$
1. Si el grado de $P(x)$ es mayor o igual al grado de $Q(x)$, efectuamos primero la división polinómica:
   $$\frac{P(x)}{Q(x)} = C(x) + \frac{R(x)}{Q(x)} \implies \int \frac{P(x)}{Q(x)} dx = \int C(x) dx + \int \frac{R(x)}{Q(x)} dx$$
   donde la integral del polinomio de cociente $C(x)$ es inmediata y la integral del resto $R(x)/Q(x)$ es una fracción propia (grado del numerador estrictamente menor que el denominador).
2. Si la fracción es propia, aplicamos el **Método de Descomposición en Fracciones Simples** factorizando el denominador $Q(x) = 0$:
   - **Caso A (Raíces reales simples)**: Si $Q(x) = (x-x_1)(x-x_2)\dots(x-x_n)$, descomponemos:
     $$\frac{R(x)}{Q(x)} = \frac{A_1}{x-x_1} + \frac{A_2}{x-x_2} + \dots + \frac{A_n}{x-x_n}$$
   - **Caso B (Raíces reales múltiples)**: Si una raíz $a$ tiene multiplicidad $n$, la descomposición para ese término incluye todas las potencias del factor:
     $$\frac{A_1}{x-a} + \frac{A_2}{(x-a)^2} + \dots + \frac{A_n}{(x-a)^n}$$
   - **Caso C (Raíces complejas conjugadas)**: Aporta términos cuadráticos irreducibles de la forma $\frac{Ax + B}{x^2 + px + q}$.

### 4.3 Resolución Paso a Paso del Ejemplo del Texto
**Ejemplo**: Evaluar la integral propia:
$$\int \frac{3x - 1}{x^2 + 2x + 5} dx$$

* **Completación del cuadrado en el denominador**:
  $$x^2 + 2x + 5 = (x + 1)^2 + 4$$
* **Descomponer el numerador** en términos del factor $(x+1)$ y su derivada:
  $$3x - 1 = 3(x + 1) - 4$$
* **Separar las integrales**:
  $$\int \frac{3x-1}{x^2+2x+5} dx = \int \frac{3(x+1) - 4}{(x+1)^2+4} dx = 3 \int \frac{x+1}{(x+1)^2+4} dx - 4 \int \frac{dx}{(x+1)^2+4}$$
* **Evaluar la primera integral** (logarítmica, haciendo $u = (x+1)^2+4 \implies du = 2(x+1)dx$):
  $$3 \int \frac{x+1}{(x+1)^2+4} dx = \frac{3}{2} \int \frac{2(x+1)}{(x+1)^2+4} dx = \frac{3}{2} \ln\left( (x+1)^2 + 4 \right) = \frac{3}{2} \ln(x^2 + 2x + 5) + C_1$$
* **Evaluar la segunda integral** (arcotangente, haciendo $w = \frac{x+1}{2} \implies dw = \frac{1}{2}dx$):
  $$4 \int \frac{dx}{(x+1)^2+4} = 4 \int \frac{dx}{4\left[\left(\frac{x+1}{2}\right)^2 + 1\right]} = \int \frac{2 \cdot dw}{w^2 + 1} = 2 \arctan(w) = 2 \arctan\left(\frac{x+1}{2}\right) + C_2$$
* **Resultado final**:
  $$\int \frac{3x-1}{x^2+2x+5} dx = \frac{3}{2}\ln(x^2 + 2x + 5) - 2 \arctan\left(\frac{x+1}{2}\right) + C$$ 


```python
# Resolución Simbólica con SymPy del ejemplo racional
f_rac = (3*x - 1) / (x**2 + 2*x + 5)
int_rac = sp.integrate(f_rac, x)

print("Descomposición de Integrales de Funciones Racionales:")
print(f"  Integrando: f(x) = {f_rac}")
print(f"  Resultado analítico obtenido por SymPy: {int_rac} + C")
print("  (Nota: Coincide con el resultado teórico simplificado)")
```

## 5. Integración por Sustitución Trigonométrica

Cuando el integrando contiene expresiones cuadráticas del teorema de Pitágoras, podemos eliminar la raíz utilizando identidades trigonométricas asociadas a triángulos rectángulos. Existen tres casos estándar de cambio de variable:

| Caso de la Expresión | Cambio de Variable (Sustitución) | Diferencial ($du$) | Identidad Pitagórica Usada |
| :---: | :---: | :---: | :--- |
| **Caso I**: $\sqrt{a^2 - u^2}$ | $u = a \sin z$ | $du = a \cos z dz$ | $1 - \sin^2 z = \cos^2 z$ |
| **Caso II**: $\sqrt{a^2 + u^2}$ | $u = a \tan z$ | $du = a \sec^2 z dz$ | $1 + \tan^2 z = \sec^2 z$ |
| **Caso III**: $\sqrt{u^2 - a^2}$ | $u = a \sec z$ | $du = a \sec z \tan z dz$ | $\sec^2 z - 1 = \tan^2 z$ |

### 5.2 Ejemplos de Aplicación Detallados de la Literatura

* **Ejemplo I (Caso I)**: Evaluar $\int \frac{\sqrt{16-x^2}}{x} dx$.
  Haciendo $x = 4\sin z \implies dx = 4\cos z dz$. La raíz es $\sqrt{16-16\sin^2 z} = 4\cos z$.
  $$\int \frac{4\cos z}{4\sin z} 4\cos z dz = 4 \int \frac{\cos^2 z}{\sin z} dz = 4 \int \frac{1-\sin^2 z}{\sin z} dz = 4 \int (\csc z - \sin z) dz$$
  $$= 4\ln|\csc z - \cot z| + 4\cos z + C$$
  Mediante la relación del triángulo para $\sin z = x/4$, tenemos $\csc z = 4/x$, $\cot z = \frac{\sqrt{16-x^2}}{x}$, y $\cos z = \frac{\sqrt{16-x^2}}{4}$.
  Sustituyendo de regreso:
  $$4 \ln\left| \frac{4 - \sqrt{16-x^2}}{x} \right| + \sqrt{16-x^2} + C$$

* **Ejemplo II (Caso I)**: Evaluar $\int \frac{\sqrt{9-x^2}}{x^2} dx$.
  Haciendo $x = 3\sin z \implies dx = 3\cos z dz$. La raíz es $3\cos z$.
  $$\int \frac{3\cos z}{9\sin^2 z} 3\cos z dz = \int \frac{9\cos^2 z}{9\sin^2 z} dz = \int \cot^2 z dz = \int (\csc^2 z - 1) dz = -\cot z - z + C$$
  Por el triángulo para $\sin z = x/3$, tenemos $\cot z = \frac{\sqrt{9-x^2}}{x}$, y $z = \arcsin(x/3)$.
  Sustituyendo de regreso:
  $$-\frac{\sqrt{9-x^2}}{x} - \arcsin\left(\frac{x}{3}\right) + C$$

* **Ejemplo III (Caso III)**: Evaluar $\int \frac{dx}{x^3 \sqrt{x^2 - 9}}$.
  Haciendo $x = 3\sec z \implies dx = 3\sec z \tan z dz$. La raíz es $\sqrt{9\sec^2 z - 9} = 3\tan z$.
  $$\int \frac{3\sec z \tan z}{27\sec^3 z \cdot 3\tan z} dz = \frac{1}{27} \int \frac{dz}{\sec^2 z} = \frac{1}{27} \int \cos^2 z dz = \frac{1}{54} \int (1 + \cos 2z) dz$$
  $$= \frac{1}{54} \left( z + \frac{1}{2}\sin 2z \right) + C = \frac{1}{54} \left( z + \sin z \cos z \right) + C$$
  Por el triángulo para $\sec z = x/3$, tenemos $\cos z = 3/x$, $\sin z = \frac{\sqrt{x^2-9}}{x}$, y $z = \text{arcsec}(x/3)$.
  Sustituyendo de regreso:
  $$\frac{1}{54} \text{arcsec}\left(\frac{x}{3}\right) + \frac{\sqrt{x^2 - 9}}{18x^2} + C$$


```python
# Graficación de los triángulos de referencia para los 3 Casos de Sustitución Trigonométrica
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 5))

# Caso I: u = a*sin(z) -> opuesto u, hipotenusa a, adyacente sqrt(a^2-u^2)
ax1.plot([0, 4, 4, 0], [0, 0, 3, 0], 'k-', linewidth=2)
ax1.text(2, -0.35, '$\\sqrt{a^2-u^2}$', ha='center', fontweight='bold', color='darkgreen')
ax1.text(4.2, 1.5, '$u$', va='center', fontweight='bold', color='blue')
ax1.text(2, 1.7, '$a$', va='center', fontweight='bold', color='red')
ax1.text(0.7, 0.2, '$z$', fontsize=12, color='darkred')
ax1.set_title('Caso I: $\\sqrt{a^2-u^2}$', fontweight='bold')
ax1.axis('off')
ax1.set_xlim(-0.5, 5.0)
ax1.set_ylim(-0.8, 3.5)

# Caso II: u = a*tan(z) -> opuesto u, adyacente a, hipotenusa sqrt(a^2+u^2)
ax2.plot([0, 4, 4, 0], [0, 0, 3, 0], 'k-', linewidth=2)
ax2.text(2, -0.35, '$a$', ha='center', fontweight='bold', color='darkgreen')
ax2.text(4.2, 1.5, '$u$', va='center', fontweight='bold', color='blue')
ax2.text(1.8, 1.7, '$\\sqrt{a^2+u^2}$', va='center', fontweight='bold', color='red')
ax2.text(0.7, 0.2, '$z$', fontsize=12, color='darkred')
ax2.set_title('Caso II: $\\sqrt{a^2+u^2}$', fontweight='bold')
ax2.axis('off')
ax2.set_xlim(-0.5, 5.0)
ax2.set_ylim(-0.8, 3.5)

# Caso III: u = a*sec(z) -> adyacente a, hipotenusa u, opuesto sqrt(u^2-a^2)
ax3.plot([0, 4, 4, 0], [0, 0, 3, 0], 'k-', linewidth=2)
ax3.text(2, -0.35, '$a$', ha='center', fontweight='bold', color='darkgreen')
ax3.text(4.2, 1.5, '$\\sqrt{u^2-a^2}$', va='center', fontweight='bold', color='blue')
ax3.text(2, 1.7, '$u$', va='center', fontweight='bold', color='red')
ax3.text(0.7, 0.2, '$z$', fontsize=12, color='darkred')
ax3.set_title('Caso III: $\\sqrt{u^2-a^2}$', fontweight='bold')
ax3.axis('off')
ax3.set_xlim(-0.5, 5.0)
ax3.set_ylim(-0.8, 3.5)

plt.tight_layout()
plt.show()
```

## 6. Integrales No Elementales

### 6.1 Definición de Función Elemental
Una función real de variable real es **elemental** si se puede construir mediante un número finito de operaciones aritméticas básicas (suma, resta, producto, cociente), potencias, radicación y composición aplicadas sobre las funciones base (polinomios, racionales, exponenciales, logarítmicas y trigonométricas directas e inversas).

### 6.2 El Límite de la Integrabilidad Elemental
Por el teorema de Liouville, no toda función elemental tiene una primitiva que pueda expresarse en términos de funciones elementales. Cuando esto ocurre, se dice que la función posee una **primitiva no elemental** (también llamadas integrales "no resuelvibles" en términos sencillos).

### 6.3 Catálogo Temático de Integrales No Elementales
A continuación se detallan las principales integrales no elementales catalogadas por familias:

1. **Funciones Algebraicas**:
   - Integrales elípticas de primera y segunda especie: $\int \frac{dx}{\sqrt{(1-x^2)(1-k^2x^2)}}$ y $\int \sqrt{\frac{1-k^2x^2}{1-x^2}} dx$.
   - Radicales de polinomios superiores ($n>2$): $\int \sqrt{ax^n+b} dx$ y $\int \frac{dx}{\sqrt{ax^n+b}}$.
   - Integrales binómicas de Chebyshev no racionales: $\int x^m(ax^n+b)^p dx$.
2. **Funciones Trigonométricas**:
   - Integrales de Fresnel: $\text{S}(x) = \int \sin(x^2) dx$ y $\text{C}(x) = \int \cos(x^2) dx$.
   - Seno y coseno integral: $\text{Si}(x) = \int \frac{\sin x}{x} dx$ y $\text{Ci}(x) = \int \frac{\cos x}{x} dx$.
   - Otros casos: $\int \sin(e^x) dx$, $\int \sqrt{\sin x} dx$, $\int x\tan x dx$, y $\int \frac{\sin x}{\sqrt{x}} dx$.
3. **Exponenciales y Logarítmicas**:
   - Exponenciales no lineales: $\int e^{-x^2} dx$ (que define la función error $\text{erf}(x)$) e $\int e^{1/x} dx$.
   - Integral logarítmica: $\text{li}(x) = \int \frac{dx}{\ln x}$ e $\int \sqrt{\ln x} dx$.
   - Integral de autopotenciación: $\int x^x dx$.

* **Nota Curiosa**: La integral $\int \sqrt{\tan x} dx$ **sí es elemental** (admite una primitiva algebraica mediante un cambio de variable no lineal), mientras que $\int \sqrt{\sin x} dx$ no lo es.


```python
from scipy.integrate import quad
import scipy.special as special

# Graficación de dos funciones no elementales fundamentales obtenidas por integración numérica
# 1. Función de Seno Integral: Si(x) = int_0^x (sin(t)/t) dt
# 2. Función Error: erf(x) = (2/sqrt(pi)) * int_0^x e^(-t^2) dt

x_plt = np.linspace(0.01, 15.0, 400)

# Calculamos Si(x) usando la función especial de scipy
# La función directa de seno integral de scipy es scipy.special.sici
y_si, y_ci = special.sici(x_plt)

# Calculamos erf(x) usando scipy
y_erf = special.erf(x_plt)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Graficar Seno Integral
ax1.plot(x_plt, y_si, 'crimson', linewidth=2.5, label='$Si(x) = \\int_0^x \\frac{\\sin t}{t} dt$')
ax1.axhline(np.pi / 2, color='black', linestyle=':', label='Límite al infinito $\\pi/2$')
ax1.set_title('Función Seno Integral ($Si(x)$)', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$Si(x)$')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# Graficar Función Error
ax2.plot(x_plt, y_erf, 'darkcyan', linewidth=2.5, label='$\\text{erf}(x) = \\frac{2}{\\sqrt{\\pi}} \\int_0^x e^{-t^2} dt$')
ax2.axhline(1.0, color='black', linestyle=':', label='Límite al infinito 1.0')
ax2.set_title('Función Error ($\\text{erf}(x)$)', fontweight='bold')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$\\text{erf}(x)$')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 7. Resumen de la Lección

1. **La Integral Indefinida**: Representa el conjunto de todas las infinitas primitivas $P(x) + C$ de una función. El valor de $C$ parametriza verticalmente la familia de curvas.
2. **Diferencial de una Función**: Definido como $dy = f'(x) dx$, representa el cambio de la recta tangente y sirve para formular aproximaciones lineales locales y modelar los cambios de variables.
3. **Métodos de Integración**:
   - **Cambio de variable**: Revierte la regla de la cadena.
   - **Por partes**: Revierte la regla del producto de derivadas bajo la fórmula $\int u dv = u v - \int v du$.
   - **Fracciones simples**: Resuelve funciones racionales factorizando el denominador y descomponiendo en términos lineales y cuadráticos.
   - **Sustitución trigonométrica**: Elimina términos pitagóricos mediante identidades asociadas a triángulos rectángulos.
4. **Primitivas No Elementales**: Representan funciones cuya integral no puede expresarse mediante combinaciones finitas de funciones elementales, requiriendo aproximaciones numéricas o funciones especiales para su estudio.

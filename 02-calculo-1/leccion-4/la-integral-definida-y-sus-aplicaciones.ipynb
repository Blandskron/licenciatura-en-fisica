## 1. La Integral Definida como Límite de una Suma

### 1.1 Sumas de Riemann
Consideremos una función real $f$ definida en un intervalo cerrado $[a, b]$. Para estimar el área bajo su curva, realizamos las siguientes construcciones geométricas:

1. **Partición ($P$)**: Dividimos el intervalo $[a, b]$ en $n$ subintervalos mediante un conjunto de puntos ordenados:
   $$a = x_0 < x_1 < x_2 < \dots < x_{n-1} < x_n = b$$
   El ancho del $i$-ésimo subintervalo es $\Delta x_i = x_i - x_{i-1}$.
2. **Puntos Muestra ($\bar{x}_i$)**: En cada subintervalo $[x_{i-1}, x_i]$ seleccionamos un punto arbitrario $\bar{x}_i$.
3. **Suma de Riemann ($R_P$)**: Construimos la suma de las áreas algebraicas de los rectángulos aproximados (ancho $\Delta x_i$ y altura $f(\bar{x}_i)$):
   $$R_P = \sum_{i=1}^n f(\bar{x}_i) \Delta x_i$$

### 1.2 Definición Formal de la Integral Definida
Sea $\|P\|$ la **norma de la partición**, que representa la longitud del subintervalo más largo en la partición $P$ ($||P|| = \max \{\Delta x_i\}$). La **integral definida** (o integral de Riemann) de $f$ de $a$ a $b$ se define como el límite de las sumas de Riemann cuando la norma de la partición tiende a cero:

$$\int_a^b f(x) dx = \lim_{\|P\| \to 0} \sum_{i=1}^n f(\bar{x}_i) \Delta x_i = L$$

Si este límite existe, se dice que la función $f$ es **integrable** en el intervalo cerrado $[a, b]$.
Equivalentemente, la definición establece que para todo $\epsilon > 0$ existe un $\delta > 0$ tal que para cualquier partición $P$ con $\|P\| < \delta$ y cualquier elección de puntos muestra $\bar{x}_i$, se cumple que:
$$\left| \sum_{i=1}^n f(\bar{x}_i) \Delta x_i - L \right| < \epsilon$$

Si la partición es **regular** (los subintervalos tienen la misma longitud $\Delta x = \frac{b-a}{n}$), la norma $\|P\| \to 0$ equivale a que el número de rectángulos tienda a infinito ($n \to \infty$). En este caso, el límite se simplifica a:
$$\int_a^b f(x) dx = \lim_{n \to \infty} \sum_{i=1}^n f(\bar{x}_i) \Delta x$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Función a integrar: f(x) = sin(x) en [0, pi]
def f_rie(x):
    return np.sin(x)

a, b = 0.0, np.pi
n = 8  # Número de rectángulos para visualización
dx = (b - a) / n

x_vals = np.linspace(a, b, 200)
y_vals = f_rie(x_vals)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Suma de Riemann Izquierda
ax1.plot(x_vals, y_vals, 'b-', linewidth=2.5, label='$f(x) = \\sin(x)$')
x_left = np.linspace(a, b - dx, n)
y_left = f_rie(x_left)
ax1.bar(x_left, y_left, width=dx, align='edge', alpha=0.3, edgecolor='blue', color='skyblue', label='Rectángulos (Izquierda)')
ax1.scatter(x_left, y_left, color='blue', s=40, zorder=5)
ax1.set_title(f'Suma de Riemann por la Izquierda ($n = {n}$)', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$f(x)$')
ax1.legend()

# Subplot 2: Suma de Riemann de Punto Medio
ax2.plot(x_vals, y_vals, 'b-', linewidth=2.5, label='$f(x) = \\sin(x)$')
x_mid = np.linspace(a + dx/2, b - dx/2, n)
y_mid = f_rie(x_mid)
ax2.bar(x_mid - dx/2, y_mid, width=dx, align='edge', alpha=0.3, edgecolor='green', color='lightgreen', label='Rectángulos (Punto Medio)')
ax2.scatter(x_mid, y_mid, color='darkgreen', s=40, zorder=5)
ax2.set_title(f'Suma de Riemann del Punto Medio ($n = {n}$)', fontweight='bold')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$f(x)$')
ax2.legend()

plt.tight_layout()
plt.show()
```

## 2. Propiedades de la Integral Definida e Integrabilidad

### 2.1 Definiciones Básicas
Para extender la consistencia del cálculo cuando se invierten los límites de integración o cuando coinciden:
1. **Límites de integración idénticos**: Si $f(a)$ existe, definimos:
   $$\int_a^a f(x) dx = 0$$
2. **Límites invertidos**: Si $a > b$ y $f$ es integrable en $[b, a]$, definimos:
   $$\int_a^b f(x) dx = -\int_b^a f(x) dx$$
3. **Variable Muda**: La variable de integración es una variable muda; la letra utilizada no altera el valor del límite:
   $$\int_a^b f(x) dx = \int_a^b f(t) dt = \int_a^b f(u) du$$

### 2.2 Teorema de Integrabilidad
No toda función real es integrable en un intervalo cerrado. Sin embargo, tenemos la siguiente condición suficiente:

> **Teorema**: Si la función $f$ es acotada en $[a, b]$ y continua en dicho intervalo excepto en un número finito de puntos, entonces $f$ es integrable en $[a, b]$.
>
> En particular, si $f$ es continua en todo el intervalo $[a, b]$, entonces $f$ es integrable en $[a, b]$.

Como consecuencia directa, las funciones polinómicas, senos/cosenos, y racionales (con denominador distinto de cero en el intervalo) son siempre integrables.

### 2.3 Propiedades Fundamentales de la Integral
Sean $f$ y $g$ funciones integrables en $[a, b]$ y sea $k$ una constante real:
1. **Linealidad (Factor constante)**:
   $$\int_a^b k f(x) dx = k \int_a^b f(x) dx$$
2. **Linealidad (Suma y resta)**:
   $$\int_a^b [f(x) \pm g(x)] dx = \int_a^b f(x) dx \pm \int_a^b g(x) dx$$
3. **Aditividad para Intervalos**: Para cualesquiera tres puntos reales $a, b, c$, se cumple:
   $$\int_a^c f(x) dx = \int_a^b f(x) dx + \int_b^c f(x) dx$$
4. **Preservación de Desigualdades (Comparación)**: Si $f(x) \le g(x)$ para todo $x \in [a, b]$, entonces:
   $$\int_a^b f(x) dx \le \int_a^b g(x) dx$$
5. **Propiedad de Acotación**: Si $m \le f(x) \le M$ para todo $x \in [a, b]$, entonces:
   $$m(b - a) \le \int_a^b f(x) dx \le M(b - a)$$
6. **Desigualdad del Valor Absoluto** (para $a < b$):
   $$\left| \int_a^b f(x) dx \right| \le \int_a^b |f(x)| dx$$


```python
# Verificación computacional de aditividad y valor absoluto para f(x) = sin(x)
# 1. Aditividad en intervalos: integral de 0 a 2*pi es la suma de [0, pi] y [pi, 2*pi]
# Usamos diferencias finitas acumulativas para aproximar las integrales
t_ad = np.linspace(0, 2 * np.pi, 1000)
y_ad = np.sin(t_ad)
dt_ad = t_ad[1] - t_ad[0]

int_completa = np.sum(y_ad) * dt_ad
int_seccion1 = np.sum(y_ad[t_ad <= np.pi]) * dt_ad
int_seccion2 = np.sum(y_ad[t_ad > np.pi]) * dt_ad

print("1. Verificación de Aditividad para Intervalos:")
print(f"  Integral Completa (0 a 2*pi) ~ {int_completa:.6f}")
print(f"  Suma de Secciones ([0, pi] + [pi, 2*pi]) ~ {int_seccion1 + int_seccion2:.6f}")
print(f"  Diferencia: {abs(int_completa - (int_seccion1 + int_seccion2)):.2e}")
print("-" * 65)

# 2. Desigualdad de valor absoluto: |integral f(x) dx| <= integral |f(x)| dx
abs_int = abs(int_completa)
int_abs = np.sum(np.abs(y_ad)) * dt_ad

print("2. Verificación de Desigualdad de Valor Absoluto:")
print(f"  |Integral sin(x)|  = {abs_int:.6f}")
print(f"  Integral |sin(x)|  = {int_abs:.6f}")
print(f"  ¿Se cumple |int f| <= int |f|? {abs_int <= int_abs}")
```

## 3. Cálculo por Definición y el Teorema del Valor Medio Integral

### 3.1 Cálculo Analítico por Sumas de Riemann del Ejemplo del Texto
Queremos evaluar mediante la definición de límite de sumas la integral:
$$\int_{-2}^3 (x + 3) dx$$

* **Paso 1**: Dividimos el intervalo $[-2, 3]$ en $n$ subintervalos regulares. El ancho es:
  $$\Delta x = \frac{3 - (-2)}{n} = \frac{5}{n}$$
* **Paso 2**: Tomamos los extremos derechos de cada intervalo como puntos muestra:
  $$x_i = a + i \Delta x = -2 + i \left(\frac{5}{n}\right)$$
* **Paso 3**: Evaluamos la función en cada punto muestra:
  $$f(x_i) = x_i + 3 = \left[-2 + i\left(\frac{5}{n}\right)\right] + 3 = 1 + i\left(\frac{5}{n}\right)$$
* **Paso 4**: Construimos la suma de Riemann regular:
  $$\sum_{i=1}^n f(x_i) \Delta x = \sum_{i=1}^n \left[ 1 + i\left(\frac{5}{n}\right) \right] \frac{5}{n} = \frac{5}{n} \sum_{i=1}^n 1 + \frac{25}{n^2} \sum_{i=1}^n i$$
  Recordando las identidades de sumación de potencias:
  $$\sum_{i=1}^n 1 = n, \quad \sum_{i=1}^n i = \frac{n(n+1)}{2}$$
  Sustituyendo y simplificando:
  $$\sum_{i=1}^n f(x_i) \Delta x = \frac{5}{n} (n) + \frac{25}{n^2} \frac{n(n+1)}{2} = 5 + \frac{25}{2} \left( 1 + \frac{1}{n} \right) = 5 + 12.5\left(1 + \frac{1}{n}\right)$$
* **Paso 5**: Tomamos el límite cuando $n \to \infty$:
  $$\int_{-2}^3 (x + 3) dx = \lim_{n \to \infty} \left[ 5 + 12.5 \left( 1 + \frac{1}{n} \right) \right] = 5 + 12.5 = 17.5 = \frac{35}{2}$$
  
Este resultado coincide exactamente con el área geométrica del trapecio formado bajo la recta $y = x + 3$ en el intervalo $[-2, 3]$:
$$Area = \frac{f(-2) + f(3)}{2} \cdot (3 - (-2)) = \frac{1 + 6}{2} \cdot 5 = \frac{35}{2} = 17.5$$

### 3.2 Teorema del Valor Medio del Cálculo Integral
Este teorema proporciona la base conceptual para el valor promedio de una función continua:

> **Teorema**: Si $f$ es una función continua en el intervalo cerrado $[a, b]$, entonces existe al menos un número $x_0 \in [a, b]$ tal que:
> $$\int_a^b f(x) dx = f(x_0) (b - a)$$

El valor $f(x_0)$ se define como el **valor medio** de la función $f$ en el intervalo $[a, b]$:
$$f(x_0) = \frac{1}{b - a} \int_a^b f(x) dx$$
Geométricamente, el teorema indica que el área neta bajo la curva es igual al área de un rectángulo con el mismo ancho $(b - a)$ y altura dada por el valor promedio de la función $f(x_0)$.


```python
# Graficación del ejemplo lineal y verificación del área del trapecio
def f_trap(x):
    return x + 3

a_t, b_t = -2.0, 3.0
x_trap = np.linspace(-3.0, 4.0, 200)
y_trap = f_trap(x_trap)

plt.figure(figsize=(10, 6.5))
plt.plot(x_trap, y_trap, 'b-', linewidth=2.5, label='$y = x + 3$')

# Colorear la región del trapecio
x_fill = np.linspace(a_t, b_t, 100)
plt.fill_between(x_fill, f_trap(x_fill), 0, color='lightgreen', alpha=0.3, label='Región de integración (Trapecio)')

# Marcar alturas de los lados del trapecio
plt.vlines([-2.0, 3.0], 0, [f_trap(-2.0), f_trap(3.0)], colors='darkgreen', linestyles='--')
plt.text(-2.1, f_trap(-2.0)/2, f'Base menor = {f_trap(-2.0)}', ha='right', color='darkgreen', fontweight='bold')
plt.text(3.1, f_trap(3.0)/2, f'Base mayor = {f_trap(3.0)}', ha='left', color='darkgreen', fontweight='bold')

plt.title('Cálculo por Definición e Interpretación Geométrica de Trapecio', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-3.0, 4.0)
plt.ylim(-0.5, 7.0)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. Teoremas Fundamentales del Cálculo y Regla de Barrow

Los Teoremas Fundamentales unifican el Cálculo Diferencial con el Cálculo Integral, demostrando que la derivación y la integración son operaciones inversas.

### 4.1 Primer Teorema Fundamental del Cálculo (FTC 1)
Si $f$ es una función continua en el intervalo cerrado $[a, b]$, y definimos la función de acumulación $F(x)$ para $x \in [a, b]$ por:
$$F(x) = \int_a^x f(t) dt$$
entonces $F(x)$ es continua en $[a, b]$, derivable en $(a, b)$, y su derivada es:
$$F'(x) = \frac{d}{dx} \left[ \int_a^x f(t) dt \right] = f(x)$$

* **Demostración Analítica**:
  Aplicando la definición de derivada para $F(x)$:
  $$F'(x) = \lim_{h \to 0} \frac{F(x+h) - F(x)}{h} = \lim_{h \to 0} \frac{1}{h} \left[ \int_a^{x+h} f(t) dt - \int_a^x f(t) dt \right]$$
  Por la aditividad de intervalos:
  $$F'(x) = \lim_{h \to 0} \frac{1}{h} \int_x^{x+h} f(t) dt$$
  Dado que $f$ es continua, aplicando el Teorema del Valor Medio del Cálculo Integral en el intervalo $[x, x+h]$ (de longitud $h$), existe un punto intermedio $c_h$ entre $x$ y $x+h$ tal que:
  $$\int_x^{x+h} f(t) dt = f(c_h) \cdot h$$
  Sustituyendo esta igualdad en el límite:
  $$F'(x) = \lim_{h \to 0} \frac{1}{h} [f(c_h) \cdot h] = \lim_{h \to 0} f(c_h)$$
  Dado que $c_h \in [x, x+h]$, cuando $h \to 0$, por el Teorema del Emparedado, se tiene que $c_h \to x$. Por la continuidad de $f$:
  $$\lim_{h \to 0} f(c_h) = f(x)$$
  Por lo tanto, $F'(x) = f(x)$. Queda demostrado.

### 4.2 Segundo Teorema Fundamental del Cálculo (Regla de Barrow)
Permite calcular integrales definidas de forma directa si conocemos una antiderivada de la función:

> **Teorema**: Si $f$ es continua en $[a, b]$ y $F$ es una primitiva cualquiera de $f$ en dicho intervalo (es decir, $F'(x) = f(x)$), entonces:
> $$\int_a^b f(x) dx = F(b) - F(a) = F(x) \Big|_a^b$$

### 4.3 Tabla de Integrales Inmediatas
A continuación se presentan las antiderivadas directas de las funciones elementales:

| Función $f(x)$ | Integral $\int f(x) dx$ | Condición |
| :--- | :--- | :--- |
| $x^n$ | $\frac{x^{n+1}}{n+1} + C$ | $n \neq -1$ |
| $x^{-1} = \frac{1}{x}$ | $\ln|x| + C$ | $x \neq 0$ |
| $e^x$ | $e^x + C$ | | 
| $a^x$ | $\frac{a^x}{\ln a} + C$ | $a > 0, a \neq 1$ |
| $\cos x$ | $\sin x + C$ | |
| $\sin x$ | $-\cos x + C$ | |
| $\sec^2 x = \frac{1}{\cos^2 x}$ | $\tan x + C$ | |
| $\csc^2 x = \frac{1}{\sin^2 x}$ | $-\cot x + C$ | |
| $\frac{1}{1+x^2}$ | $\arctan x + C$ | |
| $\frac{1}{\sqrt{1-x^2}}$ | $\arcsin x + C$ | |


```python
import sympy as sp

# Integración Simbólica con SymPy para verificar Barrow y acumulación
t = sp.Symbol('t')
x = sp.Symbol('x')

# 1. Definimos la función de acumulación F(x) = int_0^x cos(t) dt
f_integrand = sp.cos(t)
F_acum = sp.integrate(f_integrand, (t, 0, x))

print("1. Verificación del Primer Teorema Fundamental:")
print(f"  Función interna: f(t) = {f_integrand}")
print(f"  Función de acumulación F(x): {F_acum}")
print(f"  Derivada dF/dx: {sp.diff(F_acum, x)} (Coincide con f(x))")
print("-" * 65)

# 2. Regla de Barrow para evaluar integral de x^2 de 1 a 3
f_poly = x**2
F_poly = sp.integrate(f_poly, x) # Primitiva sin constante
val_barrow = sp.integrate(f_poly, (x, 1, 3))

print("2. Verificación de la Regla de Barrow para int_1^3 x^2 dx:")
print(f"  Primitiva F(x): {F_poly}")
print(f"  F(3) = {F_poly.subs(x, 3)}")
print(f"  F(1) = {F_poly.subs(x, 1)}")
print(f"  F(3) - F(1) = {F_poly.subs(x, 3) - F_poly.subs(x, 1)} = {val_barrow}")
```

## 5. Aplicaciones Geométricas de la Integral

### 5.1 Áreas de Recintos Planos
Si $f$ es continua y no negativa en $[a, b]$, el área de la región acotada por la gráfica de $y = f(x)$, las rectas verticales $x = a, x = b$ y el eje $x$ viene dada por:
$$A = \int_a^b f(x) dx$$

**Ejemplo de Cálculo**: Formular y evaluar el área bajo la curva de la función $f(x) = 1 + \sqrt{x}$ en el intervalo $[0, 4]$:
$$Area = \int_0^4 (1 + \sqrt{x}) dx = \left[ x + \frac{2}{3}x^{3/2} \right]_0^4 = \left(4 + \frac{2}{3}(4)^{3/2}\right) - 0 = 4 + \frac{2}{3}(8) = 4 + \frac{16}{3} = \frac{28}{3} \approx 9.33$$

### 5.2 Longitud de Arco de una Curva
La longitud de una curva plana o espacial suave parametrizada por vector $\vec{r}(t) = \langle f(t), g(t), h(t) \rangle$ para $a \le t \le b$ se calcula acumulando los diferenciales de longitud $|r'(t)|dt$:

$$L = \int_a^b |\vec{r}'(t)| dt = \int_a^b \sqrt{\left(\frac{dx}{dt}\right)^2 + \left(rac{dy}{dt}\right)^2 + \left(\frac{dz}{dt}\right)^2} dt$$

Si la curva en 2D es de la forma explícita $y = f(x)$, entonces parametrizando como $x = t, y = f(t)$ obtenemos la expresión clásica:
$$L = \int_a^b \sqrt{1 + [f'(x)]^2} dx$$

### 5.3 Volúmenes de Cuerpos Sólidos (Método de Rebanado)
Consideremos un sólido que se extiende a lo largo del eje $x$ desde $x = a$ hasta $x = b$. Si dividimos el sólido mediante planos perpendiculares al eje $x$, y determinamos que el área de la sección transversal resultante en cada punto $x$ es una función conocida $A(x)$:

El volumen diferencial de una rebanada es $dV = A(x)dx$. Integrando sobre todo el intervalo obtenemos el volumen total:
$$V = \int_a^b A(x) dx$$


```python
from mpl_toolkits.mplot3d import Axes3D

# Creamos un panel con 3 visualizaciones de las aplicaciones geométricas
fig = plt.figure(figsize=(18, 5.5))

# 1. Área bajo la curva f(x) = 1 + sqrt(x) en [0, 4]
ax1 = fig.add_subplot(131)
x_ar = np.linspace(0, 4.5, 200)
y_ar = 1 + np.sqrt(x_ar)
ax1.plot(x_ar, y_ar, 'b-', linewidth=2.5, label='$f(x) = 1 + \\sqrt{x}$')
x_fill = np.linspace(0, 4.0, 100)
ax1.fill_between(x_fill, 1 + np.sqrt(x_fill), color='skyblue', alpha=0.4, label='Área = 28/3 ~ 9.33')
ax1.set_title('Área bajo la curva $f(x) = 1 + \\sqrt{x}$', fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Longitud de Arco 3D de una hélice helicoidal: r(t) = (cos(t), sin(t), t)
ax2 = fig.add_subplot(132, projection='3d')
t_hel = np.linspace(0, 4 * np.pi, 300)
x_hel = np.cos(t_hel)
y_hel = np.sin(t_hel)
z_hel = t_hel

# Derivadas: dx/dt = -sin(t), dy/dt = cos(t), dz/dt = 1
# Magnitud: sqrt((-sin)^2 + cos^2 + 1^2) = sqrt(1 + 1) = sqrt(2)
# Longitud de 0 a 4*pi: L = int_0^(4*pi) sqrt(2) dt = 4*pi*sqrt(2) ~ 17.77
ax2.plot(x_hel, y_hel, z_hel, 'g-', linewidth=2.5, label='Hélice: $\\vec{r}(t)$')
ax2.set_title('Longitud de Arco en 3D: Hélice Hélice', fontweight='bold')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()

# 3. Rebanado de volumen de un paraboloide (sólido de revolución)
ax3 = fig.add_subplot(133)
x_v = np.linspace(0, 2.0, 100)
y_v = x_v**2  # Perfil de revolución y = x^2
ax3.plot(x_v, y_v, 'k-', linewidth=2.0, label='Perfil $y = x^2$')
ax3.fill_between(x_v, y_v, color='lightgreen', alpha=0.3, label='Sólido generado')

# Dibujar algunas rebanadas circulares de ancho dx
rebanadas_x = [0.5, 1.0, 1.5, 1.8]
for rx in rebanadas_x:
    radio = rx**2
    ax3.add_patch(plt.Rectangle((rx - 0.05, -radio), 0.1, 2*radio, 
                                color='darkgreen', alpha=0.6, label='Rebanadas dV' if rx==0.5 else ""))

ax3.set_title('Volumen por Rebanado (Sólido de Revolución)', fontweight='bold')
ax3.set_xlabel('$x$')
ax3.set_ylabel('Perfil $y$')
ax3.set_ylim(-4.2, 4.2)
ax3.legend()
ax3.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 6. Resumen de la Lección

1. **Sumas de Riemann**: Proporcionan la base de aproximación por rectángulos bajo la curva. Su límite al refinar la partición define la integral definida.
2. **Propiedades de Integración**: Se estructuran bajo reglas de linealidad (sumas y escalares), aditividad de intervalos de dominio ($\int_a^c = \int_a^b + \int_b^c$) y preservación de inecuaciones.
3. **Teorema del Valor Medio**: Garantiza que existe un punto donde la función toma su valor medio, el cual es equivalente a la altura promedio de la curva en el intervalo.
4. **Teoremas Fundamentales del Cálculo**: Establecen la relación inversa entre derivar e integrar. El FTC 1 demuestra que la derivada de la acumulación es la función original, y el FTC 2 (Barrow) simplifica la evaluación de integrales a través de primitivas.
5. **Aplicaciones**: La integral definida acumula diferenciales para resolver magnitudes geométricas de áreas en el plano, longitudes de curvas en 2D/3D y volúmenes de sólidos mediante rebanados transversales.

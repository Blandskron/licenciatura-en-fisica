## 1. Definición de Derivada e Interpretaciones

### 1.1 Definición Analítica de Derivada
El problema fundamental del cálculo diferencial es cuantificar con precisión la variación instantánea de una función. De acuerdo con Granville *et al.*, si $y = f(x)$ es una función continua, la **derivada** de $f$ con respecto a $x$ es el límite del cociente entre el incremento de la función y el incremento de la variable independiente cuando este último tiende a cero.

Formalmente, la derivada de una función $f$ en un punto $a$, denotada por $f'(a)$ (notación de Lagrange), se define como:
$$f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}$$
siempre que dicho límite exista.

Una notación alternativa (haciendo $x = a + h \implies h = x - a$) es:
$$f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$$

Se dice que $f$ es **derivable** (o diferenciable) en el punto $a$ si este límite existe. Si $y = f(x)$, la derivada se denota indistintamente por:
$$y' = \frac{dy}{dx} = \frac{d}{dx}f(x) = D_x f(x) = f'(x)$$

### 1.2 Interpretación Geométrica
Geométricamente, la derivada $f'(a)$ representa la **pendiente de la recta tangente** a la curva $y = f(x)$ en el punto $P(a, f(a))$.

Consideremos un punto cercano $Q(a+h, f(a+h))$ sobre la curva. La recta que pasa por $P$ y $Q$ es una **recta secante** cuya pendiente es:
$$m_{sec} = \frac{\Delta y}{\Delta x} = \frac{f(a+h) - f(a)}{h}$$
Conforme $h \to 0$, el punto $Q$ se desplaza a lo largo de la curva aproximándose indefinidamente a $P$. En el límite, la recta secante gira y su posición límite coincide con la **recta tangente** a la curva en $P$. Por tanto, la pendiente de la recta tangente $m_{tan}$ es:
$$m_{tan} = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h} = f'(a)$$

### 1.3 Interpretación Física (Tasas de Cambio)
En física, la derivada representa la **tasa de cambio instantánea** de una magnitud con respecto a otra.
Si un móvil se desplaza a lo largo de una recta y su posición en el tiempo $t$ viene dada por la función de posición $s(t)$:
* **Velocidad Promedio**: En el intervalo $[a, a+h]$, la velocidad media es el cociente de la distancia recorrida y el tiempo transcurrido:
  $$v_{prom} = \frac{s(a+h) - s(a)}{h}$$
* **Velocidad Instantánea**: Es la velocidad puntual en el instante $t = a$, definida como el límite de la velocidad promedio cuando el intervalo de tiempo tiende a cero:
  $$v(a) = s'(a) = \lim_{h \to 0} \frac{s(a+h) - s(a)}{h}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de estilo visual premium
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

# Función cuadrática f(x) = x^2
def f_geom(x):
    return x**2

a = 1.0
f_a = f_geom(a)

x_vals = np.linspace(0.2, 2.5, 300)
y_vals = f_geom(x_vals)

plt.figure(figsize=(10, 7.5))
plt.plot(x_vals, y_vals, 'k-', linewidth=2.5, label='Curva: $f(x) = x^2$')
plt.plot(a, f_a, 'ko', markersize=8, label='Punto de tangencia $P(1, 1)$')

# Recta Tangente en x = 1 (pendiente = 2*1 = 2)
y_tang = 2 * (x_vals - a) + f_a
plt.plot(x_vals, y_tang, 'r-', linewidth=2.0, label='Recta Tangente ($m_{tan} = 2$)')

# Dibujar un par de secantes para diferentes h
colores_sec = ['#ff7f0e', '#2ca02c', '#d62728']
h_vals = [1.0, 0.5, 0.2]

for h, col in zip(h_vals, colores_sec):
    q_x = a + h
    q_y = f_geom(q_x)
    # Pendiente de la secante
    m_sec = (q_y - f_a) / h
    y_sec = m_sec * (x_vals - a) + f_a
    plt.plot(x_vals, y_sec, color=col, linestyle='--', alpha=0.8, 
             label=f'Secante $h = {h}$ ($m_{{sec}} = {m_sec:.1f}$)')
    plt.plot(q_x, q_y, 'o', color=col, markersize=6)

plt.title('Aproximación de Rectas Secantes a la Recta Tangente ($h \\to 0$)', fontweight='bold')
plt.xlabel('Variable independiente ($x$)')
plt.ylabel('Función ($f(x)$)')
plt.xlim(0.3, 2.3)
plt.ylim(-0.2, 5.0)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 2. Cálculo de Derivadas y Reglas de Derivación

### 2.1 Derivación por Definición: Ejemplos Analíticos
Para consolidar la mecánica del límite de incrementos, desarrollamos los dos ejemplos clásicos del texto de Stewart *et al.*:

**Ejemplo 1**: Hallar la derivada de $f(x) = 5x^2 + 3x - 1$ en el punto $a = 2$.
Aplicando la definición de derivada en un punto:
$$f'(2) = \lim_{h \to 0} \frac{f(2+h) - f(2)}{h}$$
Primero evaluamos los términos:
$$f(2) = 5(2)^2 + 3(2) - 1 = 20 + 6 - 1 = 25$$
$$f(2+h) = 5(2+h)^2 + 3(2+h) - 1 = 5(4 + 4h + h^2) + 6 + 3h - 1 = 20 + 20h + 5h^2 + 5 + 3h = 25 + 23h + 5h^2$$
Sustituimos en el límite:
$$f'(2) = \lim_{h \to 0} \frac{(25 + 23h + 5h^2) - 25}{h} = \lim_{h \to 0} \frac{23h + 5h^2}{h} = \lim_{h \to 0} (23 + 5h) = 23$$
El resultado es $f'(2) = 23$.

**Ejemplo 2**: Hallar la derivada de $f(x) = \sqrt{x}$ en un punto genérico $a > 0$.
Aplicando la definición:
$$f'(a) = \lim_{h \to 0} \frac{\sqrt{a+h} - \sqrt{a}}{h}$$
Para levantar la indeterminación $0/0$, racionalizamos el numerador multiplicando y dividiendo por la expresión conjugada $(\sqrt{a+h} + \sqrt{a})$:
$$f'(a) = \lim_{h \to 0} \frac{(\sqrt{a+h} - \sqrt{a})(\sqrt{a+h} + \sqrt{a})}{h(\sqrt{a+h} + \sqrt{a})} = \lim_{h \to 0} \frac{(a+h) - a}{h(\sqrt{a+h} + \sqrt{a})} = \lim_{h \to 0} \frac{h}{h(\sqrt{a+h} + \sqrt{a})}$$
Cancelando $h$:
$$f'(a) = \lim_{h \to 0} \frac{1}{\sqrt{a+h} + \sqrt{a}} = \frac{1}{\sqrt{a} + \sqrt{a}} = \frac{1}{2\sqrt{a}}$$
Evaluando para los valores específicos requeridos:
$$f'(1) = \frac{1}{2\sqrt{1}} = \frac{1}{2}, \quad f'(4) = \frac{1}{2\sqrt{4}} = \frac{1}{4}, \quad f'(9) = \frac{1}{2\sqrt{9}} = \frac{1}{6}$$

### 2.2 Derivabilidad y Continuidad
Existe una estrecha relación jerárquica entre la existencia de la derivada y la continuidad de una función:

> **Proposición**: Si una función $f$ es derivable en un punto $a$, entonces $f$ es continua en $a$.
>
> *Demostración*: Debemos probar que $\lim_{x \to a} f(x) = f(a)$, lo cual equivale a demostrar que $\lim_{x \to a} [f(x) - f(a)] = 0$.
> Para todo $x \neq a$, podemos escribir:
> $$f(x) - f(a) = \frac{f(x) - f(a)}{x - a} \cdot (x - a)$$
> Tomando el límite en ambos lados cuando $x \to a$:
> $$\lim_{x \to a} [f(x) - f(a)] = \lim_{x \to a} \left[ \frac{f(x) - f(a)}{x - a} \cdot (x - a) \right] = \lim_{x \to a} \frac{f(x) - f(a)}{x - a} \cdot \lim_{x \to a} (x - a)$$
> Dado que la función es derivable en $a$, el primer límite es $f'(a)$ y el segundo es $0$:
> $$\lim_{x \to a} [f(x) - f(a)] = f'(a) \cdot 0 = 0 \implies \lim_{x \to a} f(x) = f(a)$$
> Por tanto, la función es continua.

* **La Recíproca es Falsa**: La continuidad es una condición necesaria pero **no suficiente** para la derivabilidad. Existen funciones continuas que no son derivables en ciertos puntos (donde la gráfica presenta "picos", "cúspides" o tangentes verticales). El ejemplo clásico es la función valor absoluto $f(x) = |x|$, que es continua en $x = 0$, pero no es derivable en dicho punto debido a que sus derivadas laterales difieren:
  $$f'_{-}(0) = -1 \quad \neq \quad f'_{+}(0) = 1$$

### 2.3 Reglas Algebraicas de Derivación
Dadas dos funciones $f$ y $g$ derivables en $a$:
1. **Regla de la Suma**:
   $$(f+g)'(a) = f'(a) + g'(a)$$
2. **Regla del Producto**:
   $$(fg)'(a) = f'(a)g(a) + f(a)g'(a)$$
3. **Regla del Cociente** (si $g(a) \neq 0$):
   $$\left(\frac{f}{g}\right)'(a) = \frac{f'(a)g(a) - f(a)g'(a)}{(g(a))^2}$$

### 2.4 Derivada de Función Compuesta: Regla de la Cadena
Si $f$ es derivable en $a$ y $g$ es derivable en $b = f(a)$, entonces la función compuesta $h = g \circ f$ (definida por $h(x) = g(f(x))$) es derivable en $a$ y su derivada es:
$$h'(a) = g'(f(a)) \cdot f'(a)$$

* **Demostración Rigurosa (Carathéodory / Cauchy)**:
  Definimos una función auxiliar $\phi(y)$ sobre el dominio de $g$:
  $$\phi(y) = \begin{cases} \frac{g(y) - g(b)}{y - b} & \text{si } y \neq b \\ g'(b) & \text{si } y = b \end{cases}$$
  Puesto que $g$ es derivable en $b$, $\lim_{y \to b} \phi(y) = g'(b) = \phi(b)$, lo que significa que $\phi$ es **continua en $b$**.
  De la definición de $\phi(y)$, para todo $y$, se cumple la igualdad:
  $$g(y) - g(b) = \phi(y)(y - b)$$
  Haciendo la sustitución $y = f(x)$ y $b = f(a)$, obtenemos:
  $$g(f(x)) - g(f(a)) = \phi(f(x))(f(x) - f(a))$$
  Dividiendo por $x - a$ para todo $x \neq a$:
  $$\frac{h(x) - h(a)}{x - a} = \phi(f(x)) \cdot \frac{f(x) - f(a)}{x - a}$$
  Tomando el límite cuando $x \to a$:
  - Dado que $f$ es derivable en $a$, es continua en $a$, por lo que $\lim_{x \to a} f(x) = f(a) = b$.
  - Como $\phi$ es continua en $b$, $\lim_{x \to a} \phi(f(x)) = \phi(b) = g'(b) = g'(f(a))$.
  - El límite del segundo factor es $f'(a)$.
  Por lo tanto:
  $$h'(a) = \lim_{x \to a} \frac{h(x) - h(a)}{x - a} = g'(f(a)) \cdot f'(a)$$
  Queda demostrado.


```python
import sympy as sp

# 1. Derivación Simbólica con SymPy
x_sym = sp.Symbol('x')
f_sym = 5*x_sym**2 + 3*x_sym - 1

# Calcular derivada
df_sym = sp.diff(f_sym, x_sym)
df_evaluated = df_sym.subs(x_sym, 2)

print("Derivación Analítica:")
print(f"  Función original: f(x) = {f_sym}")
print(f"  Derivada analítica: f'(x) = {df_sym}")
print(f"  Derivada evaluada en x = 2: f'(2) = {df_evaluated}")
print("-" * 55)

# 2. Aproximación Numérica de Diferencias Finitas Centrales
def f_num(x):
    return 5*x**2 + 3*x - 1

x_val = 2.0
h_list = [10**(-i) for i in range(1, 8)]

print("Estudio de convergencia de diferencias finitas centrales:")
print(f"{'h':<10} | {'Derivada Aprox.':<18} | {'Error Absoluto':<18}")
print("-" * 55)
for h in h_list:
    df_aprox = (f_num(x_val + h) - f_num(x_val - h)) / (2 * h)
    err = abs(df_aprox - float(df_evaluated))
    print(f"{h:<10.1e} | {df_aprox:<18.10f} | {err:<18.10e}")
```

## 3. Derivadas Sucesivas, Clase $C^n$ y Polinomios de Taylor

### 3.1 Derivadas Sucesivas e Infinitamente Derivables
Si la derivada $f'$ de una función es también derivable, su derivada se denomina **derivada segunda** y se denota por $f''$ o $\frac{d^2y}{dx^2}$.
En general, la **derivada de orden $n$** se representa como $f^{(n)}$ o $\frac{d^ny}{dx^n}$, y se define inductivamente como la derivada de la derivada de orden $n-1$:
$$f^{(n)}(x) = \left(f^{(n-1)}(x)\right)'$$

* **Clase $C^n$**: Se dice que una función es de clase $C^n$ en un intervalo $I$ (denotado por $f \in C^n(I)$) si sus primeras $n$ derivadas existen y son funciones continuas en $I$.
* **Clase $C^\infty$ (Funciones Suaves)**: Una función es infinitamente derivable (de clase $C^\infty$) si tiene derivadas continuas de todos los órdenes (ejemplos: $e^x$, $\sin(x)$, $\cos(x)$, y los polinomios).

### 3.2 Polinomios de Taylor (Aproximación Polinómica Local)
De acuerdo con el teorema de aproximación local, una función derivable $f$ en un entorno de un punto $a$ puede ser aproximada por una función polinómica de grado menor o igual a $n$ de la forma:
$$P_n(x) = \sum_{k=0}^n \frac{f^{(k)}(a)}{k!} (x - a)^k = f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \dots + \frac{f^{(n)}(a)}{n!}(x-a)^n$$

Este es el único polinomio de grado $\le n$ que comparte con $f$ el valor de la función y sus primeras $n$ derivadas en el punto $a$:
$$P_n^{(k)}(a) = f^{(k)}(a) \quad \text{para } k = 0, 1, \dots, n$$
Y satisface que el error de aproximación se desvanece más rápido que $(x-a)^n$:
$$\lim_{x \to a} \frac{f(x) - P_n(x)}{(x-a)^n} = 0$$


```python
# Graficación de Polinomios de Taylor de f(x) = cos(x) centrados en a = 0
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_cos = np.cos(x_vals)

# Polinomios de Taylor (Maclaurin)
P1 = np.ones_like(x_vals)                          # Grado 0 y 1
P2 = 1 - (x_vals**2) / 2                            # Grado 2 y 3
P4 = 1 - (x_vals**2) / 2 + (x_vals**4) / 24          # Grado 4 y 5
P6 = 1 - (x_vals**2) / 2 + (x_vals**4) / 24 - (x_vals**6) / 720  # Grado 6 y 7

plt.figure(figsize=(11, 7))
plt.plot(x_vals, y_cos, 'k-', linewidth=3.0, label='Función Exacta: $\\cos(x)$')
plt.plot(x_vals, P1, 'r--', linewidth=1.5, label='Grado 0/1: $P_1(x) = 1$')
plt.plot(x_vals, P2, 'b-.', linewidth=1.8, label='Grado 2/3: $P_2(x) = 1 - \\frac{x^2}{2}$')
plt.plot(x_vals, P4, 'g--', linewidth=2.0, label='Grado 4/5: $P_4(x)$')
plt.plot(x_vals, P6, 'm-', linewidth=2.0, label='Grado 6/7: $P_6(x)$')
plt.plot(0, 1, 'ko', markersize=8, label='Centro $a = 0$')

plt.title('Aproximación por Polinomios de Taylor para $f(x) = \\cos(x)$ en $a = 0$', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(-2.0, 2.0)
plt.legend(loc='lower center', ncol=3)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. Teoremas de Funciones Derivables y Regla de L'Hôpital

### 4.1 Teoremas del Valor Medio
Estos teoremas vinculan los valores de una función en los extremos de un intervalo con los valores de su derivada en puntos interiores:

1. **Teorema de Rolle**:
   Sea $f: [a, b] \to \mathbb{R}$ una función continua en $[a, b]$ y derivable en $(a, b)$. Si $f(a) = f(b)$, entonces existe al menos un punto $c \in (a, b)$ tal que:
   $$f'(c) = 0$$
   
2. **Teorema del Valor Medio (Lagrange)**:
   Sea $f: [a, b] \to \mathbb{R}$ una función continua en $[a, b]$ y derivable en $(a, b)$. Entonces existe al menos un punto $x_0 \in (a, b)$ tal que:
   $$f'(x_0) = \frac{f(b) - f(a)}{b - a}$$
   *Interpretación geométrica*: Existe un punto interior $x_0$ donde la recta tangente a la curva es paralela a la recta secante que conecta los extremos $(a, f(a))$ y $(b, f(b))$.
   
3. **Teorema de Cauchy**:
   Sean $f, g: [a, b] \to \mathbb{R}$ continuas en $[a, b]$ y derivables en $(a, b)$. Entonces existe al menos un punto $x_0 \in (a, b)$ tal que:
   $$[f(b) - f(a)]g'(x_0) = [g(b) - g(a)]f'(x_0)$$

### 4.2 La Regla de L'Hôpital
La Regla de L'Hôpital permite calcular límites de cocientes que presentan formas indeterminadas de tipo $0/0$ o $\infty/\infty$. Si $f$ y $g$ son derivables en un entorno de $a$ (salvo en $a$), $g'(x) \neq 0$, y $\lim_{x \to a} f(x) = \lim_{x \to a} g(x) = 0$ (o tienden a $\pm\infty$):
$$\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}$$
siempre que el límite de la derecha exista.

* **Comprobación para $x \to \infty$**:
  Podemos extender la regla a límites en el infinito mediante el cambio de variable $x = \frac{1}{t}$. Cuando $x \to \infty$, se tiene $t \to 0^+$. Por tanto:
  $$\lim_{x \to \infty} \frac{f(x)}{g(x)} = \lim_{t \to 0^+} \frac{f(1/t)}{g(1/t)}$$
  Suponiendo la indeterminación $0/0$ en $t \to 0^+$, aplicamos la regla de L'Hôpital para la variable $t$:
  $$\lim_{t \to 0^+} \frac{f(1/t)}{g(1/t)} = \lim_{t \to 0^+} \frac{\frac{d}{dt} f(1/t)}{\frac{d}{dt} g(1/t)}$$
  Por la regla de la cadena, $\frac{d}{dt} f(1/t) = -\frac{1}{t^2} f'(1/t)$ y $\frac{d}{dt} g(1/t) = -\frac{1}{t^2} g'(1/t)$. Sustituyendo estas derivadas:
  $$\lim_{t \to 0^+} \frac{-\frac{1}{t^2} f'(1/t)}{-\frac{1}{t^2} g'(1/t)} = \lim_{t \to 0^+} \frac{f'(1/t)}{g'(1/t)} = \lim_{x \to \infty} \frac{f'(x)}{g'(x)}$$
  Queda demostrado.


```python
# Graficación del Teorema del Valor Medio (Lagrange)
def f_lag(x):
    return x**3 - 3*x**2 - x + 4

a = 0.5
b = 3.0
fa = f_lag(a)
fb = f_lag(b)

# Pendiente de la secante
m_sec = (fb - fa) / (b - a)

# Buscamos el punto x_0 tal que f'(x_0) = 3*x_0^2 - 6*x_0 - 1 = m_sec
# Planteamos 3*x_0^2 - 6*x_0 - (1 + m_sec) = 0
coefs = [3.0, -6.0, -(1.0 + m_sec)]
raices = np.roots(coefs)
# Elegimos la raíz que está en el intervalo (0.5, 3.0)
x_0 = [r for r in raices if a < r < b][0]
y_0 = f_lag(x_0)

print(f"Intervalo [a, b]: [{a}, {b}]")
print(f"Pendiente secante = {m_sec:.4f}")
print(f"Punto intermedio hallado x_0 = {x_0:.4f}")
print(f"f'(x_0) = {3*x_0**2 - 6*x_0 - 1:.4f}")

x_plot = np.linspace(0.1, 3.4, 300)
y_plot = f_lag(x_plot)

plt.figure(figsize=(10, 7.5))
plt.plot(x_plot, y_plot, 'b-', linewidth=2.5, label='$f(x) = x^3 - 3x^2 - x + 4$')

# Recta secante
plt.plot([a, b], [fa, fb], 'o--', color='orange', linewidth=2.0, label='Recta Secante')

# Recta tangente paralela
y_tang = m_sec * (x_plot - x_0) + y_0
plt.plot(x_plot, y_tang, 'g-', linewidth=2.0, label=f'Recta Tangente en $x_0 \\approx {x_0:.2f}$')
plt.plot(x_0, y_0, 'go', markersize=8)
plt.plot([a, b], [fa, fb], 'ro', markersize=6)

plt.title('Demostración del Teorema del Valor Medio de Lagrange', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(-6, 6)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 5. Monotonía, Extremos y Curvatura

### 5.1 Criterio de Monotonía (Primera Derivada)
La primera derivada de una función determina si es creciente o decreciente en un intervalo:
* **Teorema de la Monotonía**: Sea $f$ continua en $[a, b]$ y derivable en $(a, b)$:
  - Si $f'(x) > 0$ para todo $x \in (a, b)$, entonces $f$ es **estrictamente creciente** en $[a, b]$.
  - Si $f'(x) < 0$ para todo $x \in (a, b)$, entonces $f$ es **estrictamente decreciente** en $[a, b]$.

* **Demostración para $f'(x) > 0$**:
  Sean $x_1, x_2 \in [a, b]$ tales que $x_1 < x_2$. Por el Teorema del Valor Medio aplicado en $[x_1, x_2]$, existe un punto $c \in (x_1, x_2)$ tal que:
  $$f(x_2) - f(x_1) = f'(c)(x_2 - x_1)$$
  Por hipótesis, $f'(c) > 0$ ya que $c \in (a, b)$. Como $x_2 - x_1 > 0$, el producto de ambos términos positivos es positivo:
  $$f(x_2) - f(x_1) > 0 \implies f(x_1) < f(x_2)$$
  Esto demuestra que $f$ es estrictamente creciente.

### 5.2 Curvatura (Segunda Derivada) y Rectas Tangentes
La curvatura describe cómo se dobla la gráfica de una función:
* **Concavidad (Cóncava hacia abajo)**: La pendiente de las rectas tangentes disminuye continuamente conforme aumenta $x$. Las rectas tangentes quedan **por encima** de la gráfica de la función. Se verifica si $f''(x) < 0$.
* **Convexidad (Cóncava hacia arriba)**: La pendiente de las rectas tangentes aumenta continuamente conforme aumenta $x$. Las rectas tangentes quedan **por debajo** de la gráfica de la función. Se verifica si $f''(x) > 0$.

### 5.3 Puntos de Inflexión
Un **punto de inflexión** es un punto de la gráfica de la función donde la curva cambia de tipo de curvatura (de cóncava a convexa, o viceversa).
* **Condición Necesaria**: Si $x_i$ es un punto de inflexión y la segunda derivada existe y es continua, entonces $f''(x_i) = 0$.
* **Condición Suficiente**: Si $f''(x_i) = 0$ y $f'''(x_i) \neq 0$, entonces la gráfica tiene un punto de inflexión en $x_i$.

* **Ejemplo de Condición Necesaria pero no Suficiente**:
  Consideremos la función $f(x) = x^6$ (Fig. 12 del texto de Fernández). Su segunda derivada es $f''(x) = 30x^4$, y su tercera derivada en $x = 0$ es $f'''(0) = 0$.
  En el punto $x = 0$, se anula la segunda derivada ($f''(0) = 0$). No obstante, $f''(x) = 30x^4 \ge 0$ para todo $x \in \mathbb{R}$. Dado que la segunda derivada es siempre positiva (o cero), la función es siempre convexa y no cambia de curvatura en $0$. Por lo tanto, $x=0$ **no es un punto de inflexión**, confirmando que $f''(x) = 0$ no es condición suficiente por sí misma.


```python
# Gráfica de curvatura, extremos y puntos de inflexión de f(x) = x^3 - 3x
def f_curv(x):
    return x**3 - 3*x

x_vals = np.linspace(-2.2, 2.2, 300)
y_vals = f_curv(x_vals)

plt.figure(figsize=(10, 7.5))
plt.plot(x_vals, y_vals, 'k-', linewidth=2.5, label='$f(x) = x^3 - 3x$')

# Sombrear regiones de concavidad y convexidad
plt.fill_between(x_vals, y_vals, 0, where=(x_vals < 0), color='coral', alpha=0.15, label='Concavidad ($f'' < 0$)')
plt.fill_between(x_vals, y_vals, 0, where=(x_vals > 0), color='skyblue', alpha=0.15, label='Convexidad ($f'' > 0$)')

# Puntos críticos y de inflexión
plt.plot(-1.0, 2.0, 'ro', markersize=8, label='Máximo Local (-1, 2)')
plt.plot(1.0, -2.0, 'bo', markersize=8, label='Mínimo Local (1, -2)')
plt.plot(0.0, 0.0, 'go', markersize=9, label='Punto de Inflexión (0, 0)')

# Dibujar tangente en x = -1.5 (cóncava, queda arriba)
m_c = 3*(-1.5)**2 - 3
y_tang_c = m_c * (x_vals - (-1.5)) + f_curv(-1.5)
plt.plot(x_vals, y_tang_c, 'r:', linewidth=1.8, label='Tangente en tramo cóncavo (arriba)')

# Dibujar tangente en x = 1.5 (convexa, queda abajo)
m_v = 3*(1.5)**2 - 3
y_tang_v = m_v * (x_vals - (1.5)) + f_curv(1.5)
plt.plot(x_vals, y_tang_v, 'b:', linewidth=1.8, label='Tangente en tramo convexo (abajo)')

plt.title('Estudio de Monotonía, Extremos y Concavidad', fontweight='bold')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-4.5, 4.5)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 6. Problemas de Optimización

El cálculo diferencial proporciona un método directo para resolver problemas prácticos de optimización (maximizar o minimizar una función bajo restricciones físicas o geométricas).

### 6.1 Problema 1: Vallado de Área Máxima (Triángulo + Rectángulo)
Se dispone de una tela metálica de 100 metros de longitud para vallar una región formada por un triángulo equilátero de lado $x$ y un rectángulo de base $x$ y altura $y$. ¿Cuáles son los valores de $x$ e $y$ para que el área total sea máxima?

* **Modelado del Área**:
  - Área del rectángulo: $A_R = x \cdot y$
  - Altura del triángulo: $h = \frac{\sqrt{3}}{2}x$
  - Área del triángulo: $A_T = \frac{\sqrt{3}}{4}x^2$
  - Área total a maximizar:
    $$A(x, y) = \frac{\sqrt{3}}{4}x^2 + x \cdot y$$

* **Condición de Ligadura (Perímetro exterior)**:
  El perímetro vallado consta de 3 lados del rectángulo (base $x$ y dos alturas $y$) y 2 lados del triángulo equilátero ($x$ cada uno).
  $$P = 3x + 2y = 100 \implies y = 50 - \frac{3}{2}x$$

* **Función Área en una Variable**:
  $$A(x) = 50x + \left(\frac{\sqrt{3}}{4} - \frac{3}{2}\right)x^2 \approx 50x - 1.06698x^2$$

* **Maximización (Criterio de la Primera Derivada)**:
  $$A'(x) = 50 + \left(\frac{\sqrt{3} - 6}{2}\right)x = 0 \implies x = \frac{100(6 + \sqrt{3})}{33} \approx 23.44 \text{ metros}$$

* **Verificación y Altura $y$**:
  $$A''(x) = \frac{\sqrt{3} - 6}{2} \approx -2.134 < 0 \quad (\text{Máximo Absolute})$$
  $$y = 50 - \frac{50(6 + \sqrt{3})}{11} = \frac{50(5 - \sqrt{3})}{11} \approx 14.85 \text{ metros}$$

### 6.2 Problema 2: Volumen Máximo de un Prisma Recto de Base Cuadrada
De todos los prismas rectos de base cuadrada tales que el perímetro de una cara lateral es de 30 cm, hallar las dimensiones de aquel que tenga volumen máximo.

* **Modelado del Volumen**:
  $$V(x, y) = x^2 \cdot y$$
* **Condición de Ligadura** (Perímetro de una cara lateral rectangular de lados $x$ e $y$):
  $$P = 2x + 2y = 30 \implies y = 15 - x$$
* **Función Volumen en una Variable**:
  $$V(x) = x^2(15 - x) = 15x^2 - x^3$$
* **Maximización**:
  $$V'(x) = 30x - 3x^2 = 3x(10 - x) = 0 \implies x = 10 \text{ cm} \quad (x = 0 \text{ se descarta})$$
* **Verificación y Dimensiones**:
  $$V''(10) = 30 - 60 = -30 < 0 \quad (\text{Máximo})$$
  Dimensiones: $10 \times 10 \times 5 \text{ cm}$ (volumen máximo = $500 \text{ cm}^3$).


```python
# Graficación de los problemas de optimización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6.5))

# 1. Problema de la cerca: A(x) = 50x + (sqrt(3)/4 - 1.5)*x^2
x_cerca = np.linspace(0, 100/3, 200)
y_cerca = 50 * x_cerca + (np.sqrt(3)/4 - 1.5) * x_cerca**2
x_opt_cerca = 100 * (6 + np.sqrt(3)) / 33
y_opt_cerca = 50 * x_opt_cerca + (np.sqrt(3)/4 - 1.5) * x_opt_cerca**2

ax1.plot(x_cerca, y_cerca, 'crimson', linewidth=2.5, label='Área $A(x)$')
ax1.plot(x_opt_cerca, y_opt_cerca, 'ko', markersize=8, 
         label=f'Máximo: $x \\approx {x_opt_cerca:.2f}$ m\n$A \\approx {y_opt_cerca:.2f}$ m$^2$')
ax1.set_title('Optimización de Área: Triángulo + Rectángulo', fontweight='bold')
ax1.set_xlabel('Ancho de la figura ($x$, metros)')
ax1.set_ylabel('Área total ($A$, metros cuadrados)')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Problema del prisma: V(x) = 15x^2 - x^3
x_prisma = np.linspace(0, 15, 200)
y_prisma = 15 * x_prisma**2 - x_prisma**3
x_opt_prisma = 10.0
y_opt_prisma = 500.0

ax2.plot(x_prisma, y_prisma, 'darkcyan', linewidth=2.5, label='Volumen $V(x)$')
ax2.plot(x_opt_prisma, y_opt_prisma, 'ko', markersize=8, 
         label=f'Máximo: $x = 10.0$ cm\n$V = 500.0$ cm$^3$')
ax2.set_title('Optimización de Volumen: Prisma Recto', fontweight='bold')
ax2.set_xlabel('Lado de la base ($x$, cm)')
ax2.set_ylabel('Volumen del prisma ($V$, cm$^3$)')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 7. Resumen de la Lección

1. **La Derivada**: Se define formalmente como el límite del cociente de incrementos cuando la variación de la variable independiente tiende a cero. Representa la pendiente de la recta tangente (geometría) y la tasa de cambio instantánea (física, ej. velocidad).
2. **Derivabilidad y Continuidad**: Toda función derivable es continua, pero existen funciones continuas que no son derivables debido a la presencia de picos o derivadas laterales distintas (como $|x|$ en $x=0$).
3. **Reglas Algebraicas**: Simplifican el cálculo de derivadas para sumas, productos, cocientes y composiciones de funciones (Regla de la Cadena).
4. **Polinomios de Taylor**: Extienden el concepto de aproximación lineal (tangente) a aproximaciones de orden superior $n$ usando derivadas sucesivas, permitiendo modelar funciones no lineales localmente en torno a un punto.
5. **Teoremas del Valor Medio**: Los teoremas de Rolle, Lagrange (TVM) y Cauchy establecen relaciones fundamentales entre la variación promedio de una función y su derivada instantánea en puntos interiores de un intervalo.
6. **Regla de L'Hôpital**: Permite resolver indeterminaciones del tipo $0/0$ y $\infty/\infty$ mediante el cálculo de las derivadas de las funciones del numerador y denominador.
7. **Estudio de Curvas**: La primera derivada determina los intervalos de crecimiento y extremos relativos; la segunda derivada determina la concavidad/convexidad; y la tercera derivada ayuda a ratificar puntos de inflexión.
8. **Optimización**: Utiliza el cálculo diferencial para modelar y resolver problemas reales que exigen encontrar el valor óptimo (máximo o mínimo) de una variable bajo restricciones geométricas o físicas.

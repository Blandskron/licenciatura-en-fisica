## 1. Campo Eléctrico de Distribuciones Continuas

Si dividimos un objeto cargado en elementos infinitesimales de carga $dq$, el campo eléctrico diferencial $d\vec{E}$ que produce en un punto de observación $P$ es:
$$d\vec{E} = k \frac{dq}{r^2} \hat{r} = \frac{dq}{4\pi\epsilon_0 r^2} \hat{r}$$
El campo eléctrico total es la integral sobre toda la extensión cargada:
$$\vec{E} = \int d\vec{E} = \frac{1}{4\pi\epsilon_0} \int \frac{dq}{r^2} \hat{r}$$

### 1.1 Cable Cargado Finito de Longitud $L$
Consideremos un segmento de cable recto de longitud $L$ alineado sobre el eje $x$ desde $x = -L/2$ hasta $x = L/2$, con una densidad de carga lineal uniforme $\lambda$. Calcularemos el campo eléctrico en un punto $P$ sobre el eje de simetría $z$ a una altura $z$.

Por simetría, para cada elemento de carga $dq = \lambda dx$ situado en $x$, existe un elemento correspondiente en $-x$. Las componentes horizontales del campo ($E_x$) se cancelan mutuamente. Las componentes normales ($E_z$) se suman:
$$dE_z = dE \cos\theta = \frac{\lambda dx}{4\pi\epsilon_0 r^2} \cos\theta$$
Donde $r = \sqrt{x^2 + z^2}$ y $\cos\theta = \frac{z}{r} = \frac{z}{\sqrt{x^2 + z^2}}$. Sustituyendo las variables, la integral queda:
$$E_z = 2 \int_{0}^{L/2} \frac{\lambda z dx}{4\pi\epsilon_0 (x^2 + z^2)^{3/2}} = \frac{2\lambda z}{4\pi\epsilon_0} \int_{0}^{L/2} \frac{dx}{(x^2 + z^2)^{3/2}}$$

Al integrar (lo cual validaremos más adelante con SymPy), obtenemos:
$$E_z = \frac{\lambda L}{4\pi\epsilon_0 z \sqrt{z^2 + (L/2)^2}}$$

**Casos Límite:**
1. **Carga puntual ($z \gg L$):** Si nos alejamos mucho del cable, la raíz se aproxima a $z$ y el campo tiende a:
   $$E_z \approx \frac{\lambda L}{4\pi\epsilon_0 z^2} = \frac{Q}{4\pi\epsilon_0 z^2}$$
   Donde $Q = \lambda L$ es la carga total, comportándose como una carga puntual.
2. **Cable infinito ($L \gg z$):** Si el cable es muy largo comparado con la distancia de medición, el término $(L/2)^2$ domina dentro de la raíz:
   $$E_z \approx \frac{\lambda L}{4\pi\epsilon_0 z (L/2)} = \frac{2\lambda}{4\pi\epsilon_0 z}$$

### 1.2 Disco Cargado Finito de Radio $R$
Consideremos un disco plano de radio $R$ cargado uniformemente con densidad superficial $\sigma$ en el plano $xy$. Calcularemos el campo en un punto $P$ sobre el eje axial $z$.

Dividimos el disco en anillos concéntricos de radio radial $r$ y ancho $dr$. El elemento de área es $dA = 2\pi r dr$, por lo que $dq = \sigma (2\pi r dr)$. Por simetría radial, las componentes de campo paralelas al disco se anulan, quedando solo la componente axial $E_z$:
$$dE_z = dE \cos\theta = \frac{dq}{4\pi\epsilon_0 d^2} \cos\theta = \frac{\sigma (2\pi r dr)}{4\pi\epsilon_0 (r^2 + z^2)} \frac{z}{\sqrt{r^2 + z^2}}$$
$$E_z = \frac{\sigma z}{2\epsilon_0} \int_{0}^{R} \frac{r dr}{(r^2 + z^2)^{3/2}}$$

Resolviendo la integral, obtenemos para $z > 0$:
$$E_z = \frac{\sigma}{2\epsilon_0} \left( 1 - \frac{z}{\sqrt{R^2 + z^2}} \right)$$

**Casos Límite:**
1. **Plano infinito ($R \gg z$):** Si el radio del disco es muy grande en comparación con la distancia axial, el término $z/\sqrt{R^2+z^2} \to 0$, resultando en un campo constante:
   $$E_z \approx \frac{\sigma}{2\epsilon_0}$$
2. **Carga puntual ($z \gg R$):** Desarrollando mediante serie de Taylor el término $(1 + R^2/z^2)^{-1/2}$, se puede demostrar que el campo tiende a:
   $$E_z \approx \frac{Q}{4\pi\epsilon_0 z^2}$$
   Donde $Q = \sigma \pi R^2$ es la carga total del disco.


## 2. Flujo de Campo Eléctrico y Ley de Gauss

### 2.1 Flujo Eléctrico ($\Phi$)
El flujo eléctrico representa la cantidad de líneas de campo que atraviesan una superficie determinada. Para una superficie plana de área $S$ en un campo uniforme $\vec{E}$:
$$\Phi = \vec{E} \cdot \vec{S} = E S \cos\theta$$
Donde $\vec{S} = S\hat{n}$ es el vector de área orientado normal a la superficie y $\theta$ es el ángulo entre el campo y la normal.

Para superficies curvas o campos no uniformes, el flujo se define de forma infinitesimal y se integra:
$$\Phi = \int_S \vec{E} \cdot d\vec{S}$$

### 2.2 Ley de Gauss
**Forma Integral:** El flujo eléctrico neto a través de cualquier superficie cerrada (superficie gaussiana) es proporcional a la carga neta encerrada $Q_{enc}$ en su interior:
$$\oint_S \vec{E} \cdot d\vec{S} = \frac{Q_{enc}}{\epsilon_0}$$

**Forma Diferencial:** Aplicando el teorema de la divergencia de Gauss, podemos relacionar el campo local con la densidad de carga volumétrica $\rho$ en cada punto del espacio:
$$\nabla \cdot \vec{E} = \frac{\rho}{\epsilon_0}$$


## 3. Aplicaciones de la Ley de Gauss

### 3.1 Esfera Aislante de Carga Uniforme
Consideremos una esfera de radio $R$ con una densidad volumétrica de carga uniforme $\rho_0$. La carga total es $Q = \frac{4}{3}\pi R^3 \rho_0$. Por simetría esférica, el campo apunta en dirección radial y su magnitud depende de la distancia radial $r$.

1. **Exterior ($r > R$):** Tomamos una superficie gaussiana esférica de radio $r$. La carga encerrada es la carga total $Q$:
   $$E(4\pi r^2) = \frac{Q}{\epsilon_0} \implies E_{fuera}(r) = \frac{Q}{4\pi\epsilon_0 r^2} = \frac{\rho_0 R^3}{3\epsilon_0 r^2}$$
   El campo exterior es idéntico al de una carga puntual.

2. **Interior ($r < R$):** La superficie gaussiana de radio $r$ solo encierra una porción de la carga, calculada a partir del volumen interior:
   $$Q_{enc} = \rho_0 V_{int} = \rho_0 \left(\frac{4}{3}\pi r^3\right) = Q \frac{r^3}{R^3}$$
   Aplicando la Ley de Gauss:
   $$E(4\pi r^2) = \frac{1}{\epsilon_0} Q \frac{r^3}{R^3} \implies E_{dentro}(r) = \frac{Q r}{4\pi\epsilon_0 R^3} = \frac{\rho_0 r}{3\epsilon_0}$$
   Dentro de la esfera, la intensidad del campo aumenta de forma lineal con el radio.

### 3.2 Esfera de Carga No Uniforme ($\rho = \rho_0 r^2$)
Si la densidad varía radialmente como $\rho(r) = \rho_0 r^2$, la carga en el interior se calcula mediante integración por capas esféricas:
1. **Carga interior ($r < R$):**
   $$Q_{enc}(r) = \int_{0}^{r} \rho(r') 4\pi (r')^2 dr' = \int_{0}^{r} (\rho_0 r'^2) 4\pi r'^2 dr' = 4\pi\rho_0 \int_{0}^{r} r'^4 dr' = \frac{4}{5}\pi \rho_0 r^5$$
   Por Ley de Gauss, $E_{dentro}(4\pi r^2) = Q_{enc}/\epsilon_0$:
   $$E_{dentro}(r) = \frac{\rho_0 r^3}{5\epsilon_0}$$
2. **Carga exterior ($r > R$):** La carga total de la esfera es $Q_T = \frac{4}{5}\pi\rho_0 R^5$, por lo que:
   $$E_{fuera}(r) = \frac{Q_T}{4\pi\epsilon_0 r^2} = \frac{\rho_0 R^5}{5\epsilon_0 r^2}$$


## 4. Discontinuidades de Campo y Conductores

### 4.1 Discontinuidad en la Componente Normal
Al cruzar una superficie cargada con densidad superficial $\sigma$, la componente normal del campo eléctrico presenta un salto o discontinuidad. Tomando una superficie gaussiana cilíndrica muy delgada (tipo pastilla) a través de la interfaz cargada, el flujo se concentra en las tapas superior e inferior:
$$\Phi = E_{n,arriba} A - E_{n,abajo} A = \frac{\sigma A}{\epsilon_0}$$
$$\Delta E_n = E_{n,arriba} - E_{n,abajo} = \frac{\sigma}{\epsilon_0}$$

### 4.2 Conductores en Equilibrio Electrostático y el Efecto de las Puntas
En un conductor cargado en equilibrio electrostático, las cargas son libres de moverse y se repelen mutuamente, acumulándose en la superficie exterior. Esto da lugar a las siguientes consecuencias:
1. **Campo interior nulo:** $\vec{E} = 0$ dentro de cualquier parte del conductor.
2. **Campo normal en la superficie:** En el exterior inmediatamente adyacente a la superficie, el campo debe ser perpendicular a ella, con una magnitud dada por:
   $$E_{superficie} = \frac{\sigma}{\epsilon_0}$$
3. **Efecto de las puntas:** La densidad superficial de carga $\sigma$ no es uniforme si el conductor tiene una forma irregular. En regiones de alta curvatura (menor radio de curvatura $R_c$), las cargas se concentran densamente, produciendo campos eléctricos sumamente intensos de magnitud:
   $$E \propto \frac{1}{R_c}$$

Este fenómeno es el principio de funcionamiento de los pararrayos y de la **Microscopía de Campo de Iones (FIM)**, donde una punta de metal extremadamente afilada (con un radio de curvatura de pocos nanómetros) genera un campo eléctrico tan fuerte que ioniza los átomos de gas circundantes, acelerándolos hacia una pantalla detectora para obtener imágenes con resolución atómica.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar estilo de gráficos
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

K_COULOMB = 8.9876e9
EPSILON_0 = 8.8542e-12

# =======================================================================
# SIMULACIÓN 1: CABLE CARGADO FINITO VS CABLE INFINITO
# =======================================================================
L = 2.0        # Longitud del cable (m)
lam = 1e-6     # Densidad lineal de carga (C/m)

z_vals = np.logspace(-1.5, 1.0, 300) # De 0.03 m a 10 m

# Magnitudes de campo eléctrico
E_finito = (lam * L) / (4 * np.pi * EPSILON_0 * z_vals * np.sqrt(z_vals**2 + (L/2)**2))
E_infinito = (2 * lam) / (4 * np.pi * EPSILON_0 * z_vals)
E_puntual = (lam * L) / (4 * np.pi * EPSILON_0 * z_vals**2)

fig, ax = plt.subplots(figsize=(9, 5))
ax.loglog(z_vals, E_finito, 'r-', lw=2.5, label='Cable Finito ($L = 2.0$ m)')
ax.loglog(z_vals, E_infinito, 'b--', lw=1.8, label=r'Modelo de Cable Infinito ($E \propto 1/z$)')
ax.loglog(z_vals, E_puntual, 'g:', lw=1.8, label=r'Modelo de Carga Puntual ($E \propto 1/z^2$)')

ax.set_title("Campo Eléctrico E(z) de un Cable Finito vs. Cable Infinito", fontsize=12, fontweight='bold')
ax.set_xlabel("Distancia z en el eje de simetría (m)")
ax.set_ylabel("Magnitud del Campo E (N/C)")
ax.grid(True, which="both", ls="--", alpha=0.5)
ax.legend(frameon=True)
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: DISCO CARGADO VS PLANO INFINITO
# =======================================================================
R_disco = 1.5   # Radio del disco (m)
sigma = 1e-6    # Densidad superficial de carga (C/m^2)

z_grid = np.linspace(0.01, 6.0, 400)

# Calcular campos
E_disco = (sigma / (2 * EPSILON_0)) * (1.0 - z_grid / np.sqrt(R_disco**2 + z_grid**2))
E_plano = np.ones_like(z_grid) * (sigma / (2 * EPSILON_0))
Q_total = sigma * np.pi * R_disco**2
E_puntual_disco = Q_total / (4 * np.pi * EPSILON_0 * z_grid**2)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(z_grid, E_disco, 'crimson', lw=2.5, label='Disco Cargado ($R = 1.5$ m)')
ax.plot(z_grid, E_plano, 'navy', lw=1.8, ls='--', label=r'Plano Infinito ($E = \sigma / 2\epsilon_0$)')
ax.plot(z_grid, E_puntual_disco, 'forestgreen', lw=1.8, ls=':', label=r'Límite Carga Puntual ($E \propto 1/z^2$)')

ax.set_title("Campo Eléctrico Axial E(z) de un Disco vs. Plano Infinito", fontsize=12, fontweight='bold')
ax.set_xlabel("Distancia axial z (m)")
ax.set_ylabel("Magnitud del Campo E (N/C)")
ax.set_ylim(0, 1.2 * (sigma / (2 * EPSILON_0)))
ax.grid(True, ls="--", alpha=0.5)
ax.legend(frameon=True)
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 3: COMPARATIVA DE CAMPOS EN ESFERAS
# =======================================================================
R_esfera = 1.0   # Radio de la esfera (m)
Q_carga = 1e-6   # Carga neta total (C)

r_vals = np.linspace(0.001, 3.0, 500)

# 1. Esfera conductora en equilibrio (Campo interior cero, exterior puntual)
E_cond = np.zeros_like(r_vals)
E_cond[r_vals >= R_esfera] = K_COULOMB * Q_carga / r_vals[r_vals >= R_esfera]**2

# 2. Esfera aislante con carga uniforme (Interior lineal, exterior puntual)
E_unif = np.zeros_like(r_vals)
mask_int = r_vals < R_esfera
mask_ext = r_vals >= R_esfera
E_unif[mask_int] = K_COULOMB * Q_carga * r_vals[mask_int] / (R_esfera**3)
E_unif[mask_ext] = K_COULOMB * Q_carga / r_vals[mask_ext]**2

# 3. Esfera aislante con carga no uniforme (rho = rho_0 * r^2)
# E_int = K * Q * r^3 / R^5
E_nounif = np.zeros_like(r_vals)
E_nounif[mask_int] = K_COULOMB * Q_carga * r_vals[mask_int]**3 / (R_esfera**5)
E_nounif[mask_ext] = K_COULOMB * Q_carga / r_vals[mask_ext]**2

# Graficar
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(r_vals, E_cond, 'royalblue', lw=2.5, label='Esfera Conductora (Salto en $r=R$)')
ax.plot(r_vals, E_unif, 'forestgreen', lw=2.2, ls='--', label=r'Esfera Aislante Uniforme ($E \propto r$ interior)')
ax.plot(r_vals, E_nounif, 'crimson', lw=2.2, ls=':', label=r'Esfera No Uniforme $\rho \propto r^2$ ($E \propto r^3$ interior)')

ax.axvline(R_esfera, color='darkgrey', ls='-.', label='Límite de la Esfera ($r = R$)')
ax.set_title("Perfil del Campo Eléctrico E(r) para Diferentes Distribuciones Esféricas", fontsize=12, fontweight='bold')
ax.set_xlabel("Distancia radial r (m)")
ax.set_ylabel("Magnitud del Campo E (N/C)")
ax.grid(True, ls="--", alpha=0.5)
ax.legend(frameon=True)
plt.tight_layout()
plt.show()
```

## 5. Verificación Analítica con SymPy

Utilizaremos SymPy para resolver formalmente las integrales de campo que definimos al inicio del cuaderno, confirmando simbólicamente la exactitud de nuestras deducciones analíticas manuales.

1. **Integral para el Cable Finito:**
   $$\int_{0}^{L/2} \frac{dx}{(x^2 + z^2)^{3/2}} = \frac{L/2}{z^2 \sqrt{z^2 + (L/2)^2}}$$
2. **Integral para el Disco Cargado:**
   $$\int_{0}^{R} \frac{r dr}{(r^2 + z^2)^{3/2}} = \frac{1}{z} - \frac{1}{\sqrt{R^2 + z^2}}$$


```python
import sympy as sp

# Definir símbolos reales positivos
x, r, z, L, R = sp.symbols('x r z L R', real=True, positive=True)

# 1. Integración simbólica para el Cable Finito
expr_cable = 1 / (x**2 + z**2)**sp.Rational(3, 2)
int_cable = sp.integrate(expr_cable, (x, 0, L/2))
int_cable_simp = sp.simplify(int_cable)

# Expresión analítica teórica manual para el cable (el término dentro de la integral)
teor_cable = (L / 2) / (z**2 * sp.sqrt(z**2 + (L/2)**2))
diff_cable = sp.simplify(int_cable_simp - teor_cable)

# 2. Integración simbólica para el Disco Cargado
expr_disco = r / (r**2 + z**2)**sp.Rational(3, 2)
int_disco = sp.integrate(expr_disco, (r, 0, R))
int_disco_simp = sp.simplify(int_disco)

# Expresión analítica teórica manual del disco
teor_disco = 1/z - 1/sp.sqrt(R**2 + z**2)
diff_disco = sp.simplify(int_disco_simp - teor_disco)

print("======================================================================")
print("VERIFICACIÓN DE INTEGRALES DE CAMPO ELÉCTRICO CON SYMPY")
print("======================================================================")
print(f"Integral Cable Calculada:  {int_cable_simp}")
print(f"Fórmula Cable Esperada:    {teor_cable}")
print(f"Diferencia (Cable):        {diff_cable} (Esperado: 0)\n")

print(f"Integral Disco Calculada:  {int_disco_simp}")
print(f"Fórmula Disco Esperada:    {teor_disco}")
print(f"Diferencia (Disco):        {diff_disco} (Esperado: 0)")

assert diff_cable == 0 and diff_disco == 0, "Error: ¡Las integrales de verificación fallaron!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Las integrales coinciden analíticamente.")
print("======================================================================")
```

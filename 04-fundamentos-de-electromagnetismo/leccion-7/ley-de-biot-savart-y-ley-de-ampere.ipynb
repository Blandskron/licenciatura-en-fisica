## 1. Campo Magnético de una Carga Puntual en Movimiento

Una carga puntual en reposo solo produce un campo eléctrico $\vec{E}$. Sin embargo, si la carga $q$ se mueve con velocidad constante $\vec{v}$, genera adicionalmente un campo magnético $\vec{B}$ en el espacio circundante.

El campo magnético $\vec{B}$ creado por una carga móvil en un punto de campo (determinado por el vector de posición $\vec{r}$ desde la carga) viene dado por:
$$\vec{B} = \frac{\mu_0}{4\pi} \frac{q \vec{v} \times \hat{r}}{r^2}$$
Donde:
* $\hat{r} = \vec{r} / r$ es el vector unitario que apunta del punto fuente (posición de la carga) al punto de campo.
* $\mu_0 = 4\pi \times 10^{-7}\text{ T} \cdot \text{m/A}$ es la **permeabilidad del espacio libre** (o del vacío).

La magnitud del campo es:
$$B = \frac{\mu_0}{4\pi} \frac{|q| v \sin\phi}{r^2}$$
Donde $\phi$ es el ángulo entre la velocidad $\vec{v}$ y el vector radial $\vec{r}$. Las líneas de campo son círculos concéntricos en planos perpendiculares a la línea de movimiento de la carga.


## 2. Campo Magnético Creado por Corrientes: Ley de Biot-Savart

El principio de superposición aplica a los campos magnéticos. Si subdividimos un conductor continuo portador de corriente $I$ en segmentos diferenciales de longitud $d\vec{l}$, podemos estimar el campo magnético total integrando la contribución de cada elemento.

La contribución diferencial de campo $d\vec{B}$ generada por un elemento de corriente $I d\vec{l}$ es la **Ley de Biot-Savart**:
$$d\vec{B} = \frac{\mu_0}{4\pi} \frac{I d\vec{l} \times \hat{r}}{r^2}$$
Su magnitud es:
$$dB = \frac{\mu_0}{4\pi} \frac{I dl \sin\phi}{r^2}$$
Donde $\phi$ es el ángulo entre la dirección del segmento de corriente $d\vec{l}$ y la línea que apunta del segmento al punto de campo.

### 2.1 Aplicación: Ejemplo 1 de Segmento Corto
Consideremos un alambre que porta una corriente constante $I = 125\text{ A}$. Queremos calcular el campo $dB$ generado por un segmento corto de $dl = 1\text{ cm} = 0.01\text{ m}$ a una distancia de $r = 1.2\text{ m}$ en dos puntos distintos:

* **Punto P1 (a 90° del segmento, $\phi = 90^\circ$):**
  $$dB = 10^{-7} \cdot \frac{125 \cdot 0.01 \cdot \sin(90^\circ)}{1.2^2} \approx 8.68 \times 10^{-8}\text{ T}$$
  (El libro aproxima este resultado a $8.7 \times 10^{-8}\text{ T}$).
* **Punto P2 (a 30° del segmento, $\phi = 30^\circ$):**
  $$dB = 10^{-7} \cdot \frac{125 \cdot 0.01 \cdot \sin(30^\circ)}{1.2^2} \approx 4.34 \times 10^{-8}\text{ T}$$
  (El libro aproxima este resultado a $4.3 \times 10^{-8}\text{ T}$).

En ambos casos, la regla de la mano derecha indica que el campo entra perpendicularmente al plano que contiene al alambre y al radiovector.


## 3. Campo Magnético de un Conductor Recto Largo

Para calcular el campo total creado por un alambre recto de longitud finita $2a$ portador de corriente $I$, colocamos el alambre a lo largo del eje $y$ (de $y = -a$ a $y = a$) y evaluamos el campo a una distancia perpendicular $x$ en el eje $x$:

El elemento diferencial $d\vec{l} = dy \hat{j}$. La distancia es $r = \sqrt{x^2 + y^2}$ y $\sin\phi = \sin(\pi - \phi) = x / \sqrt{x^2+y^2}$. Sustituyendo en Biot-Savart:
$$dB = \frac{\mu_0 I}{4\pi} \frac{x dy}{(x^2+y^2)^{3/2}}$$
Integrando de $-a$ a $a$:
$$B = \frac{\mu_0 I x}{4\pi} \int_{-a}^{a} \frac{dy}{(x^2+y^2)^{3/2}}$$
Mediante la sustitución trigonométrica $y = x \tan\theta$, $dy = x \sec^2\theta d\theta$, la integral indefinida se simplifica a:
$$\int \frac{dy}{(x^2+y^2)^{3/2}} = \frac{y}{x^2 \sqrt{x^2+y^2}}$$
Evaluando entre los límites de integración, obtenemos:
$$B = \frac{\mu_0 I}{4\pi} \frac{2a}{x \sqrt{x^2 + a^2}}$$

### 3.1 Límite para Conductor Infinitamente Largo
Si la longitud del alambre es mucho mayor que la distancia axial ($a \gg x$), entonces $\sqrt{x^2 + a^2} \approx a$, lo que simplifica la expresión a:
$$B = \frac{\mu_0 I}{2\pi x}$$
El campo magnético decae inversamente con la distancia $x$ al conductor, y sus líneas son círculos concéntricos al alambre.

### 3.2 Aplicación: Ejemplo de la Pared del Edificio
Un alambre vertical conduce una corriente constante $I = 25\text{ A}$ hacia arriba. Calculamos el campo a una distancia de $10\text{ cm} = 0.1\text{ m}$ a la derecha del alambre:
$$B = \frac{(4\pi \times 10^{-7}) \cdot 25}{2\pi \cdot 0.1} = 5.0 \times 10^{-5}\text{ T}$$
Por la regla de la mano derecha, a la derecha del alambre el campo magnético entra en la página.


## 4. Fuerza Magnética entre Conductores Paralelos

Dado que un conductor con corriente genera un campo magnético y experimenta fuerzas magnéticas al estar inmerso en uno, dos conductores portadores de corriente paralelos interaccionan entre sí mediante fuerzas de origen magnético.

Consideremos dos conductores paralelos separados por una distancia $d$, portando corrientes $I_1$ e $I_2$. El campo creado por el conductor 1 a la distancia del conductor 2 es $B_1 = \mu_0 I_1 / 2\pi d$. La fuerza ejercida por este campo sobre una longitud $L$ del conductor 2 es:
$$F_2 = I_2 L B_1 = I_2 L \left(\frac{\mu_0 I_1}{2\pi d}\right)$$
La **fuerza por unidad de longitud ($F/L$)** mutua es:
$$\frac{F}{L} = \frac{\mu_0 I_1 I_2}{2\pi d}$$
* **Sentidos de la Fuerza:** Corrientes en el **mismo sentido se atraen**; corrientes en **sentidos opuestos se repelen**.
* **Definición del Ampere (SI):** Un Ampere es la corriente invariable que, al circular por dos conductores paralelos de longitud infinita en el vacío separados por 1 metro, produce una fuerza mutua de exactamente $2 \times 10^{-7}\text{ N}$ por cada metro de longitud.

### 4.1 Ejemplo 2: Equilibrio de Suspensión Magnética
Un alambre horizontal superior porta una corriente constante $I_1 = 80\text{ A}$. Se coloca un segundo alambre paralelo debajo a una distancia de $d = 20\text{ cm} = 0.2\text{ m}$. El alambre inferior tiene una densidad lineal de masa de $\lambda = 0.12\text{ g/m} = 1.2 \times 10^{-4}\text{ kg/m}$. Queremos determinar la corriente $I_2$ necesaria para mantenerlo suspendido contra la gravedad:

Para la suspensión, la fuerza de atracción magnética hacia arriba debe igualar la fuerza de gravedad hacia abajo:
$$\frac{F_B}{L} = \frac{F_g}{L} \implies \frac{\mu_0 I_1 I_2}{2\pi d} = \lambda g \implies I_2 = \frac{2\pi d \lambda g}{\mu_0 I_1}$$
Sustituyendo los valores:
$$I_2 = \frac{2\pi \cdot 0.2 \cdot (1.2 \times 10^{-4}) \cdot 9.8}{(4\pi \times 10^{-7}) \cdot 80} = 14.7\text{ A}$$
Para que sea una fuerza de atracción, la corriente en el alambre inferior debe fluir en la **misma dirección** que en el alambre superior.


## 5. Campo de Espira y Bobina de Corriente: Análisis y Corrección de Errata

### 5.1 Magnitud de Campo Axial
Para una espira circular de radio $a$ portadora de corriente $I$ colocada en el plano $yz$, evaluamos el campo en el eje de simetría $x$. Por la simetría de la espira, las componentes transversales del campo se cancelan mutuamente, quedando únicamente la componente axial:
$$B_x = \frac{\mu_0 I a^2}{2(x^2 + a^2)^{3/2}}$$
Para una bobina compacta formada por $N$ espiras circulares:
$$B_x = \frac{N \mu_0 I a^2}{2(x^2 + a^2)^{3/2}}$$
En el centro de la espira ($x=0$), el campo máximo es:
$$B_{máx} = \frac{\mu_0 N I}{2a}$$

### 5.2 Análisis y Corrección de la Errata Algebraica (Fórmula 13, pág 15)

> [!IMPORTANT]
> **Errata en la Relación de Campo con el Momento Dipolar:**
> El libro guía escribe la relación del campo axial de la bobina en función del momento dipolar magnético ($\mu = N I A = N I \pi a^2$) como:
> $$B = \frac{\mu_0 \mu}{2(x^2+a^2)^{3/2}}$$
> * **Corrección Física:** Esta fórmula es matemáticamente incorrecta ya que omite el factor $\pi$ en el denominador. Al expresar la bobina en términos de su momento dipolar, despejamos el producto del área por las vueltas: $N I a^2 = \mu / \pi$. Sustituyendo esto en la ecuación de campo axial, la expresión matemáticamente correcta es:
>   $$B_x = \frac{\mu_0 \mu}{2\pi(x^2+a^2)^{3/2}}$$
>   A grandes distancias ($x \gg a$), esta fórmula reduce correctamente al campo clásico de caída dipolar a un factor $1/x^3$:
>   $$B_x \approx \frac{\mu_0 \mu}{2\pi x^3}$$
>   Corregiremos esta errata en las deducciones y en los modelos computacionales.


## 6. Ley de Gauss para el Magnetismo

El flujo magnético $\Phi_B$ a través de una superficie abierta se define como:
$$\Phi_B = \int \vec{B} \cdot d\vec{A}$$
La unidad de flujo magnético en el SI es el **Weber (Wb)** ($1\text{ Wb} = 1\text{ T} \cdot \text{m}^2$).

### 6.1 Ley de Gauss Integral
Dado que las líneas de campo magnético son lazos cerrados continuos (no existen fuentes puntuales o monopolos magnéticos aislados donde las líneas se originen o terminen), cualquier línea de campo que entra en un volumen cerrado debe salir de él. Por tanto, el flujo magnético neto a través de cualquier **superficie cerrada** es siempre igual a cero:
$$\oint \vec{B} \cdot d\vec{A} = 0$$
Esta es la Ley de Gauss para el Magnetismo. Su contraparte diferencial (ecuación de Maxwell) es $\nabla \cdot \vec{B} = 0$, que postula rigurosamente la inexistencia de monopolos magnéticos.


## 7. Ley de Ampère

La Ley de Biot-Savart es la ley fundamental para calcular el campo de cualquier corriente. Sin embargo, para distribuciones de corriente con alto grado de simetría geométrica, la **Ley de Ampère** resulta mucho más simple.

### 7.1 Formulación de la Ley de Ampère
La integral de línea cerrada del producto escalar del campo magnético y el elemento diferencial de línea $\vec{B} \cdot d\vec{l}$ a lo largo de cualquier trayectoria cerrada (espira amperiana) es proporcional a la corriente neta encerrada por la trayectoria:
$$\oint \vec{B} \cdot d\vec{l} = \mu_0 I_{enc}$$

### 7.2 Regla de Signos de la Corriente Encerrada
Para definir el signo de las corrientes en el miembro derecho, usamos la regla de la mano derecha:
1. Se curvan los dedos de la mano derecha en el sentido en que se recorre la integral a lo largo de la trayectoria cerrada.
2. El pulgar extendido apunta en la dirección de las corrientes consideradas positivas. Las corrientes que fluyen en dirección opuesta se consideran negativas.

Si tenemos tres corrientes en conductores paralelos, dos hacia arriba ($I_1, I_3$) y una hacia abajo ($I_2$), y recorremos la espira en sentido antihorario de modo que el pulgar apunte hacia arriba:
$$I_{enc} = I_1 - I_2 + I_3$$

### 7.3 Aplicación a Conductor Recto Infinito
Por simetría, para un alambre largo con corriente $I$, el campo magnético es tangente a una espira circular concéntrica de radio $r$ y tiene el mismo valor en todos sus puntos. Evaluando la integral cerrada:
$$\oint \vec{B} \cdot d\vec{l} = \oint B dl = B \oint dl = B (2\pi r)$$
Por la Ley de Ampère:
$$B (2\pi r) = \mu_0 I \implies B = \frac{\mu_0 I}{2\pi r}$$
Esto demuestra de forma sumamente sencilla y directa la expresión que deducimos anteriormente mediante Biot-Savart e integrales complejas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: LEY DE BIOT-SAVART INTEGRADA EN 3D PARA UNA ESPIRA
# =======================================================================
R_espira = 0.1       # Radio de la espira circular (m)
I_espira = 10.0      # Corriente que circula (A)
mu_0 = 4.0 * np.pi * 1e-7

# Definir rejilla espacial en el plano XZ (plano transversal, Y=0)
x_grid = np.linspace(-0.25, 0.25, 35)
z_grid = np.linspace(-0.25, 0.25, 35)
X_mesh, Z_mesh = np.meshgrid(x_grid, z_grid)

Bx_mesh = np.zeros_like(X_mesh)
Bz_mesh = np.zeros_like(X_mesh)

# Discretización de la espira en N elementos diferenciales de corriente
theta_mesh = np.linspace(0, 2.0 * np.pi, 250)
d_theta = theta_mesh[1] - theta_mesh[0]

for th in theta_mesh:
    # Posición de la fuente diferencial dL en el plano XY (Z=0)
    x_s = R_espira * np.cos(th)
    y_s = R_espira * np.sin(th)
    z_s = 0.0
    
    # Vector diferencial dL = [-R*sin(th), R*cos(th), 0] * d_theta
    dl_x = -R_espira * np.sin(th) * d_theta
    dl_y = R_espira * np.cos(th) * d_theta
    dl_z = 0.0
    
    # Sumar contribuciones a cada punto de la grilla de campo
    for r_idx in range(X_mesh.shape[0]):
        for c_idx in range(X_mesh.shape[1]):
            x_p = X_mesh[r_idx, c_idx]
            y_p = 0.0 # Plano transversal Y=0
            z_p = Z_mesh[r_idx, c_idx]
            
            # Vector de posición relativa r
            rx = x_p - x_s
            ry = y_p - y_s
            rz = z_p - z_s
            r_len = np.sqrt(rx**2 + ry**2 + rz**2)
            
            # Evitar singularidades en la sección transversal del cable
            if r_len > 0.015:
                # Producto vectorial dL x r
                cross_x = dl_y * rz - dl_z * ry
                cross_z = dl_x * ry - dl_y * rx
                
                # Ley de Biot-Savart
                Bx_mesh[r_idx, c_idx] += (mu_0 * I_espira / (4.0 * np.pi)) * cross_x / (r_len**3)
                Bz_mesh[r_idx, c_idx] += (mu_0 * I_espira / (4.0 * np.pi)) * cross_z / (r_len**3)

B_magnitude = np.sqrt(Bx_mesh**2 + Bz_mesh**2)

fig, ax = plt.subplots(figsize=(8.5, 7.5))
strm = ax.streamplot(X_mesh, Z_mesh, Bx_mesh, Bz_mesh, color=B_magnitude*1e6, 
                    cmap='viridis', linewidth=1.5, density=1.35)
fig.colorbar(strm.lines, ax=ax, label=r'Magnitud del Campo B ($\mu$T)')

# Dibujar las secciones transversales de la espira de corriente en el plano XZ
ax.scatter([-R_espira, R_espira], [0, 0], color='red', s=100, zorder=5, label='Sección Transversal de la Espira')
ax.text(-R_espira - 0.04, 0.01, r'$\otimes$ I (entra)', color='red', fontsize=11, fontweight='bold')
ax.text(R_espira + 0.01, 0.01, r'$\odot$ I (sale)', color='red', fontsize=11, fontweight='bold')

ax.set_title("Líneas de Campo Magnético en el Plano Transversal (Plano XZ)", fontweight='bold')
ax.set_xlabel("Eje X (m)")
ax.set_ylabel("Eje Z (m)")
ax.set_xlim(-0.25, 0.25)
ax.set_ylim(-0.25, 0.25)
ax.grid(True, ls='--', alpha=0.5)
ax.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: PERFIL DE CAMPO AXIAL DE BOBINA VS. DIPOLO Y ERROR
# =======================================================================
N_vueltas = 30     # Número de espiras (Ejemplo 5)
I_corriente = 5.0  # Corriente (A)
a_radio = 0.05     # Radio de bobina (m)
mu_0_val = 4.0 * np.pi * 1e-7

# Momento dipolar magnético corregido
mu_dipolar = N_vueltas * I_corriente * np.pi * (a_radio**2)

x_axis = np.linspace(-0.35, 0.35, 600)

# Modelo exacto en el eje axial de la bobina
B_exact_axial = (N_vueltas * mu_0_val * I_corriente * (a_radio**2)) / (2.0 * (x_axis**2 + a_radio**2)**1.5)

# Modelo aproximado dipolar corregido (evitando división por cero)
B_dipole_axial = (mu_0_val * mu_dipolar) / (2.0 * np.pi * np.maximum(np.abs(x_axis), 0.015)**3)

error_pct_rel = np.abs(B_exact_axial - B_dipole_axial) / B_exact_axial * 100.0

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9.5, 8.5))

# Panel 1: Perfiles axiales
ax1.plot(x_axis, B_exact_axial * 1e6, 'royalblue', lw=2.5, label='Modelo Exacto de Bobina')
ax1.plot(x_axis, B_dipole_axial * 1e6, 'darkorange', lw=1.8, ls='--', label='Modelo de Dipolo Magnético')
ax1.set_title("Campo Magnético Axial $B_x(x)$ de una Bobina vs. Dipolo", fontweight='bold')
ax1.set_xlabel("Distancia axial x (m)")
ax1.set_ylabel(r"Campo Axial $B_x$ (\\mu T)")
ax1.set_ylim(0, np.max(B_exact_axial)*1.1e6)
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Panel 2: Error relativo logarítmico para x > a
x_far_field = x_axis[x_axis > a_radio]
err_far_field = error_pct_rel[x_axis > a_radio]
ax2.plot(x_far_field, err_far_field, 'crimson', lw=2.2, label='Error Relativo (%)')
ax2.set_yscale('log')
ax2.set_title("Error Relativo Porcentual de la Aproximación Dipolar ($x > a$)", fontweight='bold')
ax2.set_xlabel("Distancia axial x (m)")
ax2.set_ylabel("Error (%) En Escala Logarítmica")
ax2.grid(True, which='both', ls='--', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 3: CURVA DE SUSPENSIÓN MAGNÉTICA DE CONDUCTORES
# =======================================================================
I_sup = 80.0        # Corriente en el alambre superior (A) (Ejemplo 2)
lambda_kg_m = 1.2e-4 # Densidad lineal: 0.12 g/m
grav_acc = 9.8      # Gravedad (m/s^2)
mu_0_const = 4.0 * np.pi * 1e-7

F_g_per_L = lambda_kg_m * grav_acc # Fuerza de gravedad por longitud
d_range_cm = np.linspace(4.0, 45.0, 400)
d_range_m = d_range_cm / 100.0

# Corriente I2 de equilibrio en suspensión para cada distancia d
I2_equilibrium = (2.0 * np.pi * d_range_m * F_g_per_L) / (mu_0_const * I_sup)

fig, ax = plt.subplots(figsize=(8.5, 5))
ax.plot(d_range_cm, I2_equilibrium, 'darkgreen', lw=2.5, label='Corriente de Equilibrio $I_2(d)$')

# Destacar el punto del Ejemplo 2: d = 20 cm
d_target_cm = 20.0
I2_target = (2.0 * np.pi * (d_target_cm/100.0) * F_g_per_L) / (mu_0_const * I_sup)
ax.scatter([d_target_cm], [I2_target], color='red', s=100, zorder=5, 
           label=f'Ejemplo 2: d={d_target_cm} cm, I2={I2_target:.2f} A')

ax.axvline(d_target_cm, color='gray', ls=':', alpha=0.5)
ax.axhline(I2_target, color='gray', ls=':', alpha=0.5)

ax.set_title("Corriente de Suspensión $I_2$ vs. Distancia de Separación $d$", fontweight='bold')
ax.set_xlabel("Distancia de separación d (cm)")
ax.set_ylabel("Corriente Requerida $I_2$ (A)")
ax.grid(True, ls='--', alpha=0.5)
ax.legend(frameon=True)
plt.tight_layout()
plt.show()
```

Utilizaremos SymPy para evaluar simbólicamente la integral definida por sustitución trigonométrica en el cálculo del campo del conductor recto finito. Esto comprobará rigurosamente que el resultado matemático del cálculo analítico coincide exactamente con el obtenido por la integración del resolvedor de álgebra simbólica.


```python
import sympy as sp

# Definir variables simbólicas reales
x_sym, y_sym, a_sym = sp.symbols('x y a', real=True, positive=True)

# Definir integrando del conductor recto
integrando_c = x_sym / (x_sym**2 + y_sym**2)**sp.Rational(3, 2)

# Evaluar la integral definida respecto a y desde -a hasta a
integral_definida = sp.integrate(integrando_c, (y_sym, -a_sym, a_sym))
integral_simplificada = sp.simplify(integral_definida)

# Comparación con el valor esperado del libro
valor_esperado = 2.0 * a_sym / (x_sym * sp.sqrt(x_sym**2 + a_sym**2))
diff_algebraica = sp.simplify(integral_simplificada - valor_esperado)

print("======================================================================")
print("VERIFICACIÓN SIMBÓLICA DE LA INTEGRAL DEL CONDUCTOR CON SYMPY")
print("======================================================================")
print(f"Integrando:                  {integrando_c}")
print(f"Resultado de la Integral:    {integral_simplificada}")
print(f"Valor Esperado Analítico:    {valor_esperado}")
print(f"Diferencia simbólica:        {diff_algebraica} (Esperado: 0)")

assert diff_algebraica == 0, "Error: ¡La integral calculada no coincide con el valor esperado!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Se validó de forma exacta la deducción analítica.")
print("======================================================================")
```

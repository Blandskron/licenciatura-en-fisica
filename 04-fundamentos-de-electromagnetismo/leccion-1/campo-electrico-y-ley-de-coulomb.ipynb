## 1. Carga Eléctrica Puntual y Propiedades de la Materia

### 1.1 Historia del Electromagnetismo
El electromagnetismo es el resultado de siglos de investigación y descubrimientos experimentales. Los hitos más significativos en la evolución de esta disciplina incluyen:

* **Descubrimiento de la electricidad (siglo VI a.C.):** Tales de Mileto y los antiguos griegos observaron que al frotar ámbar (*elektron* en griego) con pieles de animales, este adquiría la propiedad de atraer objetos ligeros como plumas u hojas secas.
* **Descubrimiento del magnetismo (siglo XIII):** En China se descubrieron las propiedades atrayentes de la magnetita ($Fe_3O_4$) sobre el hierro, lo que condujo al desarrollo de la brújula para navegación marina.
* **Estudios Científicos Formales (siglos XVI y XVII):** William Gilbert publicó *De Magnete* (1600), acuñando el término "electricidad" y estableciendo una distinción clara entre los fenómenos eléctricos (atracción por frotamiento) y magnéticos.
* **Leyes de la Electrostática (siglo XVIII):** Charles-Augustin de Coulomb formuló en 1785 las leyes cuantitativas de la atracción y repulsión eléctrica utilizando una balanza de torsión. Henry Cavendish determinó con precisión la constante de proporcionalidad de dicha ley.
* **Relación entre Electricidad y Magnetismo (siglo XIX):** Hans Christian Oersted (1820) observó que una corriente eléctrica desvía la aguja de una brújula cercana. Poco después, Michael Faraday (1831) descubrió la inducción electromagnética al demostrar que un campo magnético variable en el tiempo induce una corriente eléctrica.
* **Ecuaciones de Maxwell (1860-1865):** James Clerk Maxwell unificó los campos eléctricos y magnéticos en un conjunto de cuatro ecuaciones diferenciales vectoriales, demostrando que la luz es una onda electromagnética.

### 1.2 Campo Eléctrico y Cargas Eléctricas Puntuales
La carga eléctrica es una propiedad intrínseca de la materia. Se presenta en dos tipos: **positiva** (portada por los protones) y **negativa** (portada por los electrones). La carga está cuantizada en múltiplos enteros de la carga elemental $e \approx 1.602 \times 10^{-19}\text{ C}$, y se mide en **culombios (C)** en el Sistema Internacional.

Una **carga eléctrica puntual** es un modelo idealizado que representa una carga concentrada en un único punto geométrico sin dimensiones (dimensión espacial cero). Este modelo es de gran utilidad cuando las dimensiones físicas de los cuerpos cargados son despreciables en comparación con las distancias de separación involucradas en el problema.

### 1.3 Conductores, Aislantes y Semiconductores
Los materiales se clasifican según su capacidad de permitir el movimiento de cargas eléctricas (electrones libres) a través de ellos:

* **Conductores:** Materiales con baja resistividad que contienen electrones de valencia débilmente unidos al núcleo, los cuales forman una "nube electrónica" móvil. Los metales como el cobre, la plata y el oro son excelentes conductores.
* **Aislantes (Dieléctricos):** Materiales con alta resistividad en los cuales los electrones están firmemente ligados a sus átomos o moléculas, impidiendo el flujo de corriente. Ejemplos comunes incluyen el vidrio, la madera seca y los plásticos.
* **Semiconductores:** Materiales que poseen una brecha de energía (band gap) estrecha entre la banda de valencia (llena) y la banda de conducción (vacía). Su conductividad puede incrementarse drásticamente mediante el aumento de temperatura o el **dopaje** (adición controlada de impurezas):
  - **Tipo N:** Dopado con átomos pentavalentes (como el fósforo) que aportan electrones libres adicionales.
  - **Tipo P:** Dopado con átomos trivalentes (como el boro) que crean "huecos" o vacantes de electrones en la banda de valencia.

#### El Efecto Joule
Cuando una corriente eléctrica fluye por un conductor, los electrones colisionan con la red cristalina del material, disipando energía en forma de calor. La potencia disipada ($P$) se expresa mediante la **Ley de Joule**:
$$P = I^2 R = \frac{V^2}{R}$$
Donde $I$ es la corriente en amperios (A), $V$ es el voltaje en voltios (V) y $R = \rho L/A$ es la resistencia eléctrica en ohmios ($\Omega$) para un conductor de longitud $L$ y sección transversal $A$.


## 2. Ley de Coulomb e Interacción entre Cargas

La Ley de Coulomb establece cuantitativamente la fuerza de atracción o repulsión entre dos cargas puntuales en reposo.

### 2.1 Formulación de la Ley de Coulomb
**Forma Escalar:** La magnitud de la fuerza electrostática $|\vec{F}|$ entre dos cargas puntuales $q_1$ y $q_2$ separadas por una distancia $r$ es directamente proporcional al producto de las cargas e inversamente proporcional al cuadrado de la distancia:
$$|\vec{F}| = k \frac{|q_1 q_2|}{r^2}$$

Donde $k$ es la constante de Coulomb. En el vacío, viene dada en el Sistema Internacional por:
$$k = \frac{1}{4\pi\epsilon_0} \approx 8.9876 \times 10^9\text{ N}\cdot\text{m}^2\cdot\text{C}^{-2}$$
Donde $\epsilon_0 \approx 8.8542 \times 10^{-12}\text{ F/m}$ representa la permitividad eléctrica del vacío. Si las cargas se encuentran en un medio material dieléctrico, se utiliza la permitividad del medio $\epsilon = \epsilon_r \epsilon_0$, donde $\epsilon_r$ es la permitividad relativa del medio.

**Forma Vectorial:** Para calcular con precisión la dirección y sentido de la fuerza, expresamos la ley vectorialmente. La fuerza $\vec{F}_{12}$ ejercida por la carga $q_1$ (en $\vec{r}_1$) sobre la carga $q_2$ (en $\vec{r}_2$) es:
$$\vec{F}_{12} = k \frac{q_1 q_2}{|\vec{r}_2 - \vec{r}_1|^3} (\vec{r}_2 - \vec{r}_1)$$

### 2.2 Interacción entre Múltiples Cargas: Principio de Superposición
La fuerza total ejercida sobre una carga de prueba $Q$ en la posición $\vec{r}_Q$ debido a un conjunto de $N$ cargas puntuales $q_i$ en las posiciones $\vec{r}_i$ es la **suma vectorial** de las fuerzas individuales ejercidas por cada carga de manera independiente:
$$\vec{F}_T = \sum_{i=1}^{N} \vec{F}_i = k Q \sum_{i=1}^{N} \frac{q_i}{|\vec{r}_Q - \vec{r}_i|^3} (\vec{r}_Q - \vec{r}_i)$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración del estilo de graficación
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# Constante de Coulomb en el vacío (SI)
K_COULOMB = 8.9876e9

# Definición de las cargas fuentes q_i: (magnitud en Coulomb, posición en metros)
cargas_fuente = [
    {"q": 5.0e-6,  "r": np.array([1.0, 1.0]),   "label": r"$q_1 = 5\ \mu$C",  "color": "royalblue"},
    {"q": -5.0e-6, "r": np.array([-1.0, 1.0]),  "label": r"$q_2 = -5\ \mu$C", "color": "crimson"},
    {"q": 3.0e-6,  "r": np.array([0.0, -1.0]),  "label": r"$q_3 = 3\ \mu$C",  "color": "forestgreen"}
]

# Carga de prueba Q y su posición r_Q
Q_prueba = 1.0e-6
r_Q = np.array([0.5, 0.2])

# Listas para almacenar las fuerzas calculadas
fuerzas_individuales = []
fuerza_neta = np.zeros(2)

# Cálculo de fuerzas aplicando el principio de superposición
for c in cargas_fuente:
    r_vec = r_Q - c["r"]
    dist = np.linalg.norm(r_vec)
    # Evitar singularidades de división por cero
    if dist < 1e-6:
        continue
    # Fuerza vectorial ejercida por c sobre Q
    F_vec = K_COULOMB * c["q"] * Q_prueba * r_vec / (dist**3)
    fuerzas_individuales.append(F_vec)
    fuerza_neta += F_vec

# Graficar la configuración y los vectores de fuerza
fig, ax = plt.subplots(figsize=(8, 8))

# Dibujar cargas fuentes
for c in cargas_fuente:
    ax.scatter(c["r"][0], c["r"][1], color=c["color"], s=150, zorder=5, edgecolors='black', label=c["label"])
    signo = "+" if c["q"] > 0 else "-"
    ax.text(c["r"][0], c["r"][1], signo, color='white', fontsize=12, ha='center', va='center', fontweight='bold', zorder=6)

# Dibujar la carga de prueba
ax.scatter(r_Q[0], r_Q[1], color='orange', s=120, zorder=5, edgecolors='black', label=r"$Q = 1\ \mu$C (Prueba)")
ax.text(r_Q[0], r_Q[1], "+", color='black', fontsize=10, ha='center', va='center', fontweight='bold', zorder=6)

# Escala de visualización de las fuerzas (1 Newton = escala unidades de pantalla)
escala_fuerza = 40.0

# Trazar vectores de fuerzas individuales
for c, F in zip(cargas_fuente, fuerzas_individuales):
    ax.quiver(r_Q[0], r_Q[1], F[0], F[1], angles='xy', scale_units='xy', scale=1.0/escala_fuerza,
              color=c["color"], alpha=0.7, width=0.005, label=f"Fuerza de {c['label'].split(' =')[0]}")

# Trazar el vector de fuerza neta
ax.quiver(r_Q[0], r_Q[1], fuerza_neta[0], fuerza_neta[1], angles='xy', scale_units='xy', scale=1.0/escala_fuerza,
          color='purple', width=0.007, zorder=7, label=rf"$\vec{{F}}_{{net}}$ = {np.linalg.norm(fuerza_neta):.3f} N")

# Detalles estéticos del gráfico
ax.set_title("Superposición de Fuerzas Electrostáticas (Ley de Coulomb)", fontsize=13, fontweight='bold')
ax.set_xlabel("Posición $x$ (m)")
ax.set_ylabel("Posición $y$ (m)")
ax.set_xlim(-2.0, 2.0)
ax.set_ylim(-2.0, 2.0)
ax.axhline(0, color='black', linewidth=0.5, alpha=0.3)
ax.axvline(0, color='black', linewidth=0.5, alpha=0.3)
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(loc='upper right', frameon=True)
plt.tight_layout()
plt.show()
```

## 3. El Campo Eléctrico y el Potencial Electrostático

### 3.1 Campo Eléctrico
El campo eléctrico $\vec{E}$ en un punto del espacio se define como la fuerza eléctrica $\vec{F}$ por unidad de carga que experimentaría una carga de prueba positiva $q$ infinitamente pequeña colocada en ese punto:
$$\vec{E} = \frac{\vec{F}}{q}$$

El campo eléctrico se expresa en unidades de **Newton por Culombio (N/C)** o equivalentemente en **Voltios por metro (V/m)**. A partir de la Ley de Coulomb, el campo generado por una carga puntual $Q$ en la posición del origen a una distancia $r$ es:
$$\vec{E} = k \frac{Q}{r^2} \hat{r} = \frac{Q}{4\pi\epsilon_0 r^2} \hat{r}$$
Donde $\hat{r}$ es el vector unitario en dirección radial hacia afuera desde la carga.

### 3.2 Potencial Eléctrico y Corrección de Errata Teórica
El potencial eléctrico $V$ en un punto representa la energía potencial electrostática por unidad de carga. Se calcula a partir de la integral de línea del campo eléctrico. 

> [!IMPORTANT]
> **Nota Crítica sobre los Límites de Integración:**
> En la literatura docente y el texto guía (página 11), a menudo se presenta la siguiente integral de manera simplificada:
> $$V = -\int_{0}^{\infty} \frac{Q}{4\pi\epsilon_0 r^2} dr$$
> Físicamente, esta formulación presenta dos inconsistencias:
> 1. La integral de $1/r^2$ desde $0$ a $\infty$ diverge matemáticamente debido al polo en $r=0$.
> 2. El signo resultante no se corresponde con la definición formal del potencial electrostático relativo al infinito.
>
> La integral matemáticamente rigurosa para calcular el potencial $V(r)$ en un punto $r$ asumiendo que el potencial es cero en el infinito ($V(\infty) = 0$) es:
> $$V(r) = -\int_{\infty}^{r} \vec{E} \cdot d\vec{r'} = -\int_{\infty}^{r} \frac{Q}{4\pi\epsilon_0 (r')^2} dr' = \left[ \frac{Q}{4\pi\epsilon_0 r'} \right]_{\infty}^{r} = \frac{Q}{4\pi\epsilon_0 r}$$

El potencial eléctrico es una magnitud escalar medida en **Voltios (V)**, donde $1\text{ V} = 1\text{ J/C}$. Esto significa que el potencial a una distancia $r$ de una carga positiva es positivo y tiende asintóticamente a cero conforme nos alejamos de la carga.

### 3.3 Principio de Superposición para Campos y Potenciales
Para un sistema de múltiples cargas puntuales $q_i$, el campo eléctrico y el potencial total en un punto de observación se calculan sumando las contribuciones individuales:
$$\vec{E}_T = \sum_{i=1}^{N} \vec{E}_i = k \sum_{i=1}^{N} \frac{q_i}{|\vec{r} - \vec{r}_i|^3} (\vec{r} - \vec{r}_i)$$
$$V_T = \sum_{i=1}^{N} V_i = k \sum_{i=1}^{N} \frac{q_i}{|\vec{r} - \vec{r}_i|}$$


```python
# Simulación 2: Visualización de Líneas de Campo Eléctrico y Potencial de un Dipolo

# Definición de las dos cargas del dipolo eléctrico
dipolo = [
    {"q": 1.0e-6,  "r": np.array([0.5, 0.0])},  # Carga positiva
    {"q": -1.0e-6, "r": np.array([-0.5, 0.0])}  # Carga negativa
]

# Rejilla de evaluación 2D
x_vals = np.linspace(-2.0, 2.0, 150)
y_vals = np.linspace(-2.0, 2.0, 150)
X, Y = np.meshgrid(x_vals, y_vals)

Ex = np.zeros(X.shape)
Ey = np.zeros(Y.shape)
V = np.zeros(X.shape)

# Cálculo del campo y el potencial en la rejilla
for c in dipolo:
    rx = X - c["r"][0]
    ry = Y - c["r"][1]
    r = np.sqrt(rx**2 + ry**2)
    # Suprimir la singularidad en el centro de las cargas
    r[r < 1e-2] = 1e-2
    
    # Suma de componentes de campo y potencial
    Ex += K_COULOMB * c["q"] * rx / (r**3)
    Ey += K_COULOMB * c["q"] * ry / (r**3)
    V += K_COULOMB * c["q"] / r

# Graficar
fig, ax = plt.subplots(figsize=(9, 8))

# Dibujar el potencial como un mapa de contorno continuo (acotando valores extremos)
V_lim = np.clip(V, -2.5e4, 2.5e4)
contour = ax.contourf(X, Y, V_lim, levels=50, cmap='RdBu_r', alpha=0.85)
cbar = fig.colorbar(contour, ax=ax)
cbar.set_label("Potencial Eléctrico $V$ (V)")

# Dibujar líneas equipotenciales específicas con trazos discontinuos
ax.contour(X, Y, V, levels=[-1.5e4, -8.0e3, -4.0e3, -2.0e3, 0.0, 2.0e3, 4.0e3, 8.0e3, 1.5e4],
           colors='black', linewidths=0.8, alpha=0.6, linestyles='dashed')

# Dibujar las líneas de flujo del campo eléctrico
ax.streamplot(X, Y, Ex, Ey, color='black', linewidth=1.0, density=1.4,
              arrowstyle='->', arrowsize=1.2, minlength=0.1)

# Graficar las posiciones de las cargas
for c in dipolo:
    color = 'red' if c["q"] > 0 else 'blue'
    signo = "+" if c["q"] > 0 else "-"
    ax.scatter(c["r"][0], c["r"][1], color=color, s=200, zorder=10, edgecolors='black')
    ax.text(c["r"][0], c["r"][1], signo, color='white', fontsize=15, ha='center', va='center', fontweight='bold', zorder=11)

ax.set_title("Líneas de Campo Eléctrico y Superficies Equipotenciales (Dipolo)", fontsize=13, fontweight='bold')
ax.set_xlabel("Posición $x$ (m)")
ax.set_ylabel("Posición $y$ (m)")
ax.set_xlim(-1.8, 1.8)
ax.set_ylim(-1.8, 1.8)
ax.grid(True, linestyle='--', alpha=0.2)
plt.tight_layout()
plt.show()
```

## 4. Distribuciones Continuas de Carga y la Ley de Gauss

### 4.1 Distribuciones Continuas de Carga
Cuando el número de partículas cargadas es sumamente grande, se modela el sistema como una distribución continua de carga. Dependiendo de la geometría, definimos tres tipos de densidad de carga:

1. **Lineal (un cable cargado):** $dq = \lambda dl$, donde $\lambda$ es la carga por unidad de longitud ($C/m$).
2. **Superficial (una placa plana):** $dq = \sigma dA$, donde $\sigma$ es la carga por unidad de área ($C/m^2$).
3. **Volumétrica (una esfera sólida):** $dq = \rho dV$, donde $\rho$ es la carga por unidad de volumen ($C/m^3$).

El campo eléctrico resultante se obtiene integrando sobre toda la distribución de carga:
$$\vec{E}(\vec{r}) = k \int \frac{dq}{|\vec{r} - \vec{r'}|^3} (\vec{r} - \vec{r'})$$

### 4.2 Campo Eléctrico en un Capacitor de Placas Paralelas: Efectos de Borde
Un capacitor está compuesto por dos conductores paralelos cargados con cargas iguales y opuestas $\pm Q$. Si las placas son infinitas y planas, el campo eléctrico en el interior es perfectamente uniforme y constante:
$$E = -\frac{\Delta V}{d}$$
Donde $\Delta V$ es la diferencia de potencial y $d$ es la distancia de separación entre las placas. Fuera de las placas, el campo ideal es nulo.

Sin embargo, para placas reales con dimensiones finitas, las líneas de campo se curvan hacia afuera en los bordes de la estructura. Este fenómeno se conoce como **efectos de borde** (fringe effects). Podemos simular este comportamiento aproximando cada placa por una fila discreta de muchas cargas puntuales distribuidas uniformemente. Al aplicar el principio de superposición sobre este conjunto, observamos cómo el campo se deforma en las fronteras del capacitor.


```python
# Simulación 3: Capacitor de Placas Paralelas y Efectos de Borde

# Parámetros geométricos
L_placa = 2.0       # Longitud de las placas (m)
d_sep = 0.6         # Distancia de separación (m)
N_particulas = 50   # Discretización de cargas por placa
q_punto = 8.0e-9    # Carga elemental en cada punto de discretización

# Posiciones y cargas de los elementos de la placa superior (cargada positivamente, y = d/2)
x_sup = np.linspace(-L_placa/2, L_placa/2, N_particulas)
y_sup = np.ones(N_particulas) * (d_sep/2)
q_sup = np.ones(N_particulas) * q_punto

# Posiciones y cargas de los elementos de la placa inferior (cargada negativamente, y = -d/2)
x_inf = np.linspace(-L_placa/2, L_placa/2, N_particulas)
y_inf = np.ones(N_particulas) * (-d_sep/2)
q_inf = np.ones(N_particulas) * (-q_punto)

# Consolidar todas las cargas discretas de las placas
x_cargas = np.concatenate([x_sup, x_inf])
y_cargas = np.concatenate([y_sup, y_inf])
q_cargas = np.concatenate([q_sup, q_inf])

# Rejilla de evaluación 2D
x_grid = np.linspace(-2.2, 2.2, 140)
y_grid = np.linspace(-1.5, 1.5, 140)
X_grid, Y_grid = np.meshgrid(x_grid, y_grid)

Ex_cap = np.zeros(X_grid.shape)
Ey_cap = np.zeros(Y_grid.shape)
V_cap = np.zeros(X_grid.shape)

# Superposición sobre la rejilla para calcular campo y potencial
for xc, yc, qc in zip(x_cargas, y_cargas, q_cargas):
    rx = X_grid - xc
    ry = Y_grid - yc
    r = np.sqrt(rx**2 + ry**2)
    r[r < 1.2e-2] = 1.2e-2  # Suprimir singularidades
    
    Ex_cap += K_COULOMB * qc * rx / (r**3)
    Ey_cap += K_COULOMB * qc * ry / (r**3)
    V_cap += K_COULOMB * qc / r

# Graficar
fig, ax = plt.subplots(figsize=(10, 7))

# Mapa de contorno continuo para el potencial eléctrico
V_cap_lim = np.clip(V_cap, -450, 450)
contour = ax.contourf(X_grid, Y_grid, V_cap_lim, levels=50, cmap='coolwarm', alpha=0.75)
cbar = fig.colorbar(contour, ax=ax)
cbar.set_label("Potencial Eléctrico $V$ (V)")

# Superficies equipotenciales específicas
ax.contour(X_grid, Y_grid, V_cap, levels=np.linspace(-300, 300, 13), colors='black', linewidths=0.6, alpha=0.4)

# Trazar líneas de campo eléctrico
ax.streamplot(X_grid, Y_grid, Ex_cap, Ey_cap, color='black', linewidth=0.9, density=1.7, 
              arrowstyle='->', arrowsize=1.0)

# Representar físicamente las placas cargadas del capacitor
ax.plot([-L_placa/2, L_placa/2], [d_sep/2, d_sep/2], color='darkred', linewidth=4.5, label='Placa Superior ($+Q$)')
ax.plot([-L_placa/2, L_placa/2], [-d_sep/2, -d_sep/2], color='darkblue', linewidth=4.5, label='Placa Inferior ($-Q$)')

# Detalles finales del gráfico
ax.set_title("Líneas de Campo y Potencial en un Capacitor: Efectos de Borde", fontsize=13, fontweight='bold')
ax.set_xlabel("Posición $x$ (m)")
ax.set_ylabel("Posición $y$ (m)")
ax.set_xlim(-2.0, 2.0)
ax.set_ylim(-1.3, 1.3)
ax.grid(True, linestyle='--', alpha=0.2)
ax.legend(loc='upper right', frameon=True)
plt.tight_layout()
plt.show()
```

### 4.3 Ley de Gauss y Jaula de Faraday
**Ley de Gauss:** Establece que el flujo neto del campo eléctrico $\Phi$ a través de cualquier superficie cerrada (llamada superficie gaussiana) es proporcional a la carga neta $Q_{enc}$ encerrada dentro de dicha superficie:
$$\Phi = \oint \vec{E} \cdot d\vec{S} = \frac{Q_{enc}}{\epsilon_0}$$
Esta ley es una de las ecuaciones de Maxwell y es una herramienta muy potente para calcular campos eléctricos cuando existen geometrías con un alto grado de simetría (esférica, cilíndrica o planar).

**Jaula de Faraday:** Es una consecuencia directa de la Ley de Gauss y del comportamiento de los materiales conductores en equilibrio electrostático. Si una superficie conductora cerrada se somete a un campo eléctrico externo, los electrones libres dentro del conductor se reorganizan rápidamente de modo que acumulan carga en la superficie exterior, generando un campo interno inducido que cancela exactamente el campo externo. Como consecuencia:
1. El campo eléctrico en el interior del conductor es estrictamente nulo ($\vec{E}_{int} = 0$).
2. El potencial eléctrico es constante en todo el volumen del conductor.
3. Cualquier carga neta en un conductor en equilibrio reside exclusivamente en su superficie exterior.

Este fenómeno tiene aplicaciones prácticas en el blindaje de dispositivos electrónicos sensibles y la seguridad contra descargas eléctricas (rayos).


## 5. Verificación Analítica con SymPy

En física matemática, el campo eléctrico electrostático es un campo conservativo, lo que significa que el trabajo realizado para mover una carga a lo largo de cualquier camino cerrado es cero. Como consecuencia, el campo eléctrico se puede obtener directamente como el **gradiente negativo del potencial eléctrico**:
$$\vec{E} = -\nabla V = -\left( \frac{\partial V}{\partial x}\hat{i} + \frac{\partial V}{\partial y}\hat{j} + \frac{\partial V}{\partial z}\hat{k} \right)$$

A continuación, utilizaremos el motor de cálculo simbólico **SymPy** para comprobar que el gradiente del potencial de una carga puntual nos devuelve exactamente el vector de campo eléctrico de la Ley de Coulomb.


```python
import sympy as sp

# Definir variables espaciales y parámetros físicos simbólicos
x, y, z = sp.symbols('x y z', real=True)
Q, eps_0 = sp.symbols('Q eps_0', positive=True)

# Distancia radial r en coordenadas cartesianas 3D
r_sym = sp.sqrt(x**2 + y**2 + z**2)

# Potencial eléctrico V de una carga puntual en el origen
V_sym = Q / (4 * sp.pi * eps_0 * r_sym)

# Calcular las componentes del gradiente de V: grad(V)
grad_V_x = sp.diff(V_sym, x)
grad_V_y = sp.diff(V_sym, y)
grad_V_z = sp.diff(V_sym, z)

# El campo eléctrico calculado como -grad(V)
E_grad_x = -grad_V_x
E_grad_y = -grad_V_y
E_grad_z = -grad_V_z

# Campo eléctrico teórico directo (Ley de Coulomb): E_i = Q * coord_i / (4 * pi * eps_0 * r^3)
E_teor_x = Q * x / (4 * sp.pi * eps_0 * r_sym**3)
E_teor_y = Q * y / (4 * sp.pi * eps_0 * r_sym**3)
E_teor_z = Q * z / (4 * sp.pi * eps_0 * r_sym**3)

# Evaluar la diferencia entre la definición teórica y el gradiente calculado
diff_x = sp.simplify(E_grad_x - E_teor_x)
diff_y = sp.simplify(E_grad_y - E_teor_y)
diff_z = sp.simplify(E_grad_z - E_teor_z)

# Imprimir los resultados de la verificación analítica
print("======================================================================")
print("VERIFICACIÓN ANALÍTICA CON SYMPY: E = -grad(V)")
print("======================================================================")
print(f"Potencial modelado V(r) = {V_sym}")
print(f"-dV/dx calculado: {sp.factor(E_grad_x)}")
print(f"E_x teórico:      {E_teor_x}\n")
print(f"Diferencia en componente X (E_grad_x - E_teor_x): {diff_x} (Esperado: 0)")
print(f"Diferencia en componente Y (E_grad_y - E_teor_y): {diff_y} (Esperado: 0)")
print(f"Diferencia en componente Z (E_grad_z - E_teor_z): {diff_z} (Esperado: 0)")

assert diff_x == 0 and diff_y == 0 and diff_z == 0, "Error: ¡Las expresiones matemáticas no coinciden!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Se ha demostrado analíticamente que E = -grad(V).")
print("======================================================================")
```

## 1. Características de la Corriente Alterna

La **corriente alterna (CA)** se define como un flujo de carga eléctrica cuya magnitud y dirección varían cíclicamente de forma sinusoidal en el tiempo. Se describe mediante la ecuación:
$$i(t) = I_{max} \sin(\omega t)$$
$$v(t) = V_{max} \sin(\omega t)$$

Donde:
* $I_{max}$ y $V_{max}$ son las amplitudes máximas de corriente y voltaje.
* $\omega = 2\pi f$ es la frecuencia angular (rad/s), siendo $f = 1/T$ la frecuencia en Hertz (Hz) y $T$ el período del ciclo.
* Frecuencias de red comunes: $50\text{ Hz}$ (Europa, $\omega \approx 314.16\text{ rad/s}$) y $60\text{ Hz}$ (América, $\omega \approx 376.99\text{ rad/s}$).

### 1.1 Valores Eficaces (rms)
Dado que el promedio de una función seno sobre un ciclo completo es cero, para medir los efectos térmicos de la corriente alterna se utiliza el valor eficaz o **rms (root-mean-square)**. Representa el valor equivalente en corriente continua que disiparía la misma potencia promedio en una resistencia:
$$I_{rms} = \frac{I_{max}}{\sqrt{2}} \approx 0.707 \cdot I_{max}$$
$$V_{rms} = \frac{V_{max}}{\sqrt{2}} \approx 0.707 \cdot V_{max}$$


## 2. Componentes Individuales en CA y Corrección de Erratas

Analizaremos el comportamiento de los tres componentes básicos al conectarse individualmente a una fuente sinusoidal de voltaje $v(t) = V_{max} \sin(\omega t)$.

### 2.1 Resistencia Pura
En un resistor, la corriente y el voltaje están estrictamente **en fase** (desfase $\phi = 0$). Cumplen la ley de Ohm instantánea:
$$i_R(t) = \frac{v_R(t)}{R} = \frac{V_{max}}{R} \sin(\omega t) = I_{max} \sin(\omega t)$$

### 2.2 Inductor Puro y Corrección de Errata de Amplitud (pág 11 - Fórmulas 10 y 12)
La caída de voltaje instantánea en un inductor es $v_L(t) = L \frac{di}{dt}$. Integrando la corriente a partir de un voltaje sinusoidal se obtiene que la corriente se **atrasa $90^\circ$ ($\pi/2$ rad)** respecto al voltaje:
$$i_L(t) = I_{max} \sin\left(\omega t - \frac{\pi}{2}\right) = -I_{max} \cos(\omega t)$$
La **reactancia inductiva** es la oposición del inductor al paso de la corriente:
$$X_L = \omega L$$

> [!IMPORTANT]
> **Corrección de Errata de Signo en Amplitud del Inductor:**
> El libro guía escribe la relación de amplitudes del inductor (Fórmulas 10 y 12) con un signo negativo:
> $$I_{max} = -\frac{V_{max}}{X_L} \quad \text{y} \quad I_{rms} = -\frac{V_{rms}}{X_L}$$
> * **Corrección Física:** Las amplitudes máximas $I_{max}$ y rms $I_{rms}$ son magnitudes escalares estrictamente reales y positivas por definición. El signo menos surge temporalmente al escribir la corriente instantánea como coseno ($i_L(t) = -I_{max} \cos(\omega t)$), pero las amplitudes máximas siempre cumplen:
>   $$I_{max} = \frac{V_{max}}{X_L} \quad \text{y} \quad I_{rms} = \frac{V_{rms}}{X_L}$$

### 2.3 Capacitor Puro y Corrección de Errata de la Fórmula 21 (pág 13)
La carga en las placas del condensador es $q(t) = C v_C(t)$. Derivando en el tiempo se obtiene la corriente, la cual se **adelanta $90^\circ$ ($\pi/2$ rad)** respecto al voltaje:
$$i_C(t) = \frac{dq}{dt} = \omega C V_{max} \cos(\omega t) = I_{max} \sin\left(\omega t + \frac{\pi}{2}\right)$$
La **reactancia capacitiva** es la oposición del capacitor en CA:
$$X_C = \frac{1}{\omega C}$$
Relación de amplitudes: $V_{max} = I_{max} X_C$.

> [!IMPORTANT]
> **Corrección de Errata Crítica en la Fórmula 21 (pág 13):**
> El libro guía escribe la ecuación temporal de voltaje en el condensador como:
> $$\Delta V_{rms} = -I_{rms} X_L \sin(\omega t)$$
> * **Corrección Física:** Esta expresión contiene tres errores graves de consistencia física:
>   1. *Inconsistencia dimensional:* Igualar un valor constante eficaz ($\Delta V_{rms}$) a una función sinusoidal temporal ($\sin(\omega t)$).
>   2. *Inconsistencia de componentes:* Utiliza la reactancia de un inductor ($X_L$) para describir el comportamiento de un condensador ($X_C$).
>   3. *Relación de desfase errónea:* La caída instantánea de voltaje del capacitor en desfase se escribe en coseno negativo o en función de fase atrasada:
>      $$v_C(t) = -V_{max} \cos(\omega t) = I_{max} X_C \sin\left(\omega t - \frac{\pi}{2}\right)$$
>      La relación rms pura entre constantes del capacitor es simplemente:
>      $$V_{rms} = I_{rms} X_C$$


## 3. El Transformador

Un **transformador** es un dispositivo eléctrico que utiliza la inducción mutua entre dos devanados aislados enrollados alrededor de un mismo núcleo ferromagnético para modificar los voltajes de corriente alterna con una eficiencia superior al $95\%$.

### 3.1 Relación de Transformación
En un transformador ideal (sin pérdidas de potencia por corrientes de Foucault o histéresis, y con acoplamiento magnético total del flujo $\Phi_B$):
* **Relación de voltajes:**
  $$\frac{v_2}{v_1} = \frac{N_2}{N_1}$$
  Donde $v_1$ es el voltaje del devanado primario ($N_1$ espiras) y $v_2$ el del secundario ($N_2$ espiras).
  * Si $N_2 > N_1$: Transformador **elevador** ($v_2 > v_1$).
  * Si $N_2 < N_1$: Transformador **reductor** ($v_2 < v_1$).
* **Relación de corrientes:** Dado que la potencia se conserva ($P_{entrada} = P_{salida} \implies v_1 I_1 = v_2 I_2$):
  $$\frac{I_1}{I_2} = \frac{N_2}{N_1} \implies I_1 = I_2 \frac{N_2}{N_1}$$

### 3.2 Acoplamiento de Impedancia
El transformador modifica la resistencia o impedancia de carga $R_L$ conectada en el secundario, de tal manera que la fuente del primario percibe una resistencia equivalente corregida:
$$R_{eq} = \frac{v_1}{I_1} = \frac{R_L}{(N_2/N_1)^2}$$
Esto es fundamental en ingeniería para el acoplamiento de impedancias y maximizar la transferencia de potencia en circuitos de audio y RF.


## 4. Fasores

Dado que en corriente alterna los voltajes y corrientes son funciones armónicas sinusoidales que tienen fases diferentes, sumarlas directamente de forma aritmética es incorrecto. Para simplificar el análisis se introducen los **fasores**.

Un **fasor** es un vector rotatorio en el plano complejo de dos dimensiones (eje real e imaginario) que gira en sentido antihorario con una velocidad angular constante $\omega$:
* **Longitud del vector:** Corresponde a la amplitud máxima ($V_{max}$ o $I_{max}$) de la señal sinusoidal.
* **Orientación (ángulo con el eje horizontal):** Representa la fase instantánea $\omega t + \theta_0$.
* **Proyección sobre el eje vertical:** Equivale al valor instantáneo físico de la magnitud en el tiempo ($v(t) = V_{max} \sin(\omega t)$).

El uso de fasores permite transformar la suma de funciones sinusoidales complejas y ecuaciones diferenciales en simples operaciones de álgebra vectorial o suma de números complejos complejos.


## 5. Circuitos RLC en Serie (LCR)

Consideremos un circuito que contiene en serie un resistor $R$, un inductor $L$ y un condensador $C$ conectados a una fuente de CA $v(t) = V_{max} \sin(\omega t)$. Al estar en serie, la corriente instantánea $i(t) = I_{max} \sin(\omega t - \phi)$ es idéntica en todos los componentes.

### 5.1 Suma Vectorial de Voltajes e Impedancia $Z$
Aplicando la ley de mallas de Kirchhoff en términos de fasores, el voltaje de la fuente es la suma de los fasores de voltaje individuales:
$$\vec{V} = \vec{V}_R + \vec{V}_L + \vec{V}_C$$
Como $\vec{V}_L$ y $\vec{V}_C$ se encuentran en la misma línea vertical pero en sentidos opuestos (desfasados $180^\circ$ entre sí), la diferencia de voltaje total reactivo es $V_L - V_C$. Por el teorema de Pitágoras:
$$V_{max} = \sqrt{V_R^2 + (V_L - V_C)^2} = \sqrt{(I_{max} R)^2 + (I_{max} X_L - I_{max} X_C)^2}$$
$$V_{max} = I_{max} \sqrt{R^2 + (X_L - X_C)^2}$$

La **impedancia** $Z$ del circuito es la oposición total al paso de corriente alterna:
$$Z = \sqrt{R^2 + (X_L - X_C)^2}$$
$$I_{max} = \frac{V_{max}}{Z}$$

### 5.2 Ángulo de Fase $\phi$
El desfase $\phi$ entre el voltaje de la fuente y la corriente del circuito se calcula como:
$$\phi = \tan^{-1}\left(\frac{X_L - X_C}{R}\right)$$
* **Circuito Inductivo ($X_L > X_C$):** $\phi > 0$. El voltaje se adelanta a la corriente (corriente retrasada).
* **Circuito Capacitivo ($X_L < X_C$):** $\phi < 0$. La corriente se adelanta al voltaje (corriente adelantada).
* **Resonancia RLC ($X_L = X_C$):** $\phi = 0$. La impedancia es mínima ($Z = R$) y la corriente es máxima. La frecuencia de resonancia natural es:
  $$f_0 = \frac{1}{2\pi \sqrt{LC}}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: DIAGRAMA DE FASORES ROTATORIOS PARA CIRCUITO RLC SERIE
# =======================================================================
V_max = 10.0        # Amplitud de voltaje de fuente (V)
R_res = 6.0         # Resistencia (Ohms)
L_ind = 0.02        # Inductancia (H)
C_cap = 500e-6      # Capacitancia (F)
f_hz = 50.0         # Frecuencia (Hz)

omega = 2.0 * np.pi * f_hz
X_L = omega * L_ind
X_C = 1.0 / (omega * C_cap)
Z_imp = np.sqrt(R_res**2 + (X_L - X_C)**2)
I_max = V_max / Z_imp
phi_rad = np.arctan2(X_L - X_C, R_res)

# Fotografía en un instante t = 0.003 segundos
t_snap = 0.003
theta_t = omega * t_snap

# Definición de fasores como números complejos acoplados a la fase
I_fasor = I_max * np.exp(1j * theta_t)
V_R_fasor = (I_max * R_res) * np.exp(1j * theta_t)
V_L_fasor = (I_max * X_L) * np.exp(1j * (theta_t + np.pi/2)) # Adelantado 90
V_C_fasor = (I_max * X_C) * np.exp(1j * (theta_t - np.pi/2)) # Retrasado 90
V_tot_fasor = V_max * np.exp(1j * (theta_t + phi_rad))

fig, ax = plt.subplots(figsize=(8.5, 8.5))

# Trazado de fasores vectoriales usando plt.quiver
ax.quiver(0, 0, np.real(V_R_fasor), np.imag(V_R_fasor), color='darkorange', 
          angles='xy', scale_units='xy', scale=1, label=r'$\vec{V}_R$ (Resistencia en fase)')
ax.quiver(0, 0, np.real(V_L_fasor), np.imag(V_L_fasor), color='royalblue', 
          angles='xy', scale_units='xy', scale=1, label=r'$\vec{V}_L$ (Inductor, +90$^{\circ}$)')
ax.quiver(0, 0, np.real(V_C_fasor), np.imag(V_C_fasor), color='crimson', 
          angles='xy', scale_units='xy', scale=1, label=r'$\vec{V}_C$ (Capacitor, -90$^{\circ}$)')
ax.quiver(0, 0, np.real(V_tot_fasor), np.imag(V_tot_fasor), color='purple', 
          angles='xy', scale_units='xy', scale=1, lw=2.5, label=r'$\vec{V}$ (Voltaje de Fuente)')
ax.quiver(0, 0, np.real(I_fasor)*5.0, np.imag(I_fasor)*5.0, color='forestgreen', 
          angles='xy', scale_units='xy', scale=1, ls='--', alpha=0.85, label=r'$\vec{I} \times 5.0$ (Corriente)')

# Trazado auxiliar de la suma geométrica V_L + V_C
V_reactivo = V_L_fasor + V_C_fasor
ax.quiver(np.real(V_R_fasor), np.imag(V_R_fasor), np.real(V_reactivo), np.imag(V_reactivo), 
          color='gray', angles='xy', scale_units='xy', scale=1, ls=':', alpha=0.55)

ax.set_xlim(-11.5, 11.5)
ax.set_ylim(-11.5, 11.5)
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)
ax.set_title(f"Diagrama de Fasores Rotatorios en Plano Complejo ($t = {t_snap*1000:.1f}\text{{ ms}}$)", fontweight='bold')
ax.set_xlabel("Componente Real (Voltios)")
ax.set_ylabel("Componente Imaginario (Voltios)")
ax.grid(True, ls='--', alpha=0.5)
ax.legend(frameon=True, loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: CURVAS DE RESONANCIA DE IMPEDANCIA Y DESFASE EN RLC
# =======================================================================
V_max_src = 12.0    # V
R_val = 10.0        # Ohms
L_val = 0.05        # H
C_val = 10e-6       # F (10 uF)

# Frecuencia de resonancia natural f0 = 1 / (2 * pi * sqrt(LC))
f_resonance = 1.0 / (2.0 * np.pi * np.sqrt(L_val * C_val))

# Rango de barrido de frecuencia: 10 Hz a 750 Hz
f_array = np.linspace(10.0, 750.0, 800)
omega_array = 2.0 * np.pi * f_array

X_L_array = omega_array * L_val
X_C_array = 1.0 / (omega_array * C_val)
Z_array = np.sqrt(R_val**2 + (X_L_array - X_C_array)**2)
I_max_array = V_max_src / Z_array
phi_degrees = np.degrees(np.arctan2(X_L_array - X_C_array, R_val))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9.5, 9.5))

# Panel 1: Impedancia Z y Corriente máxima Imax
ax1.plot(f_array, Z_array, 'crimson', lw=2.2, label=r'Impedancia $Z(f)$ ($\Omega$)')
ax1_twin = ax1.twinx()
ax1_twin.plot(f_array, I_max_array, 'navy', lw=2.5, label='Corriente Máxima $I_{max}(f)$ (A)')
ax1.axvline(f_resonance, color='darkgreen', ls='--', lw=2.0, 
           label=f'Frecuencia de Resonancia $f_0 = {f_resonance:.1f}\text{{ Hz}}$')

ax1.set_title("Curvas de Impedancia y Corriente vs. Frecuencia (Resonancia RLC)", fontweight='bold')
ax1.set_xlabel("Frecuencia f (Hz)")
ax1.set_ylabel(r"Impedancia $Z$ ($\Omega$)", color='crimson')
ax1_twin.set_ylabel("Corriente Máxima $I_{max}$ (A)", color='navy')
ax1.grid(True, ls='--', alpha=0.5)

# Fusión de leyendas de ambos ejes de ordenadas
handles_1, labels_1 = ax1.get_legend_handles_labels()
handles_2, labels_2 = ax1_twin.get_legend_handles_labels()
ax1.legend(handles_1 + handles_2, labels_1 + labels_2, frameon=True, loc='upper right')

# Panel 2: Ángulo de desfase phi
ax2.plot(f_array, phi_degrees, 'darkorange', lw=2.5, label=r'Desfase Voltaje-Corriente $\phi$')
ax2.axvline(f_resonance, color='darkgreen', ls='--', lw=2.0)
ax2.axhline(0, color='gray', ls=':', alpha=0.7)
ax2.set_title("Ángulo de Desfase entre Voltaje y Corriente vs. Frecuencia", fontweight='bold')
ax2.set_xlabel("Frecuencia f (Hz)")
ax2.set_ylabel(r"Ángulo de Desfase $\phi$ (grados)")
ax2.set_ylim(-90, 90)
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True, loc='lower right')

plt.tight_layout()
plt.show()
```

## 7. Verificación Analítica con SymPy

Utilizaremos la teoría de números complejos en SymPy para representar simbólicamente la impedancia compleja de un circuito RLC serie:
$$\bar{Z} = R + j \left(\omega L - \frac{1}{\omega C}\right)$$
Donde $j$ es la unidad imaginaria (`sp.I`). Validaremos de forma exacta que el módulo $|Z| = \sqrt{\bar{Z} \cdot \bar{Z}^*}$ y la fase de esta expresión coinciden con las fórmulas analíticas trigonométricas deducidas mediante fasores.


```python
import sympy as sp

# Definir variables simbólicas positivas reales
R_sym, L_sym, C_sym, omega_sym = sp.symbols('R L C omega', real=True, positive=True)

# Plantear la impedancia compleja serie
Z_complex = R_sym + sp.I * (omega_sym * L_sym - 1.0 / (omega_sym * C_sym))

# Módulo o magnitud de Z
Z_mod = sp.Abs(Z_complex)

# Ángulo de fase argumental
phi_sym = sp.arg(Z_complex)

# Expresiones convencionales esperadas
Z_expected = sp.sqrt(R_sym**2 + (omega_sym * L_sym - 1.0 / (omega_sym * C_sym))**2)

# Comparación simbólica de diferencias al cuadrado
diff_magnitude_sq = sp.simplify(Z_mod**2 - Z_expected**2)

print("======================================================================")
print("VERIFICACIÓN DE IMPEDANCIA COMPLEJA CON SYMPY")
print("======================================================================")
print(f"Impedancia Compleja Serie Z_c:  {Z_complex}")
print(f"Módulo Calculado Simbólico |Z|:  {sp.simplify(Z_mod)}")
print(f"Fórmula de Módulo Esperada:     {Z_expected}")
print(f"Diferencia cuadrática:          {diff_magnitude_sq} (Esperado: 0)")
print(f"Fórmula de Fase phi (arg):      {phi_sym}")

assert diff_magnitude_sq == 0, "Error: ¡La validación del módulo falló!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! La álgebra compleja de impedancia es exacta.")
print("======================================================================")
```

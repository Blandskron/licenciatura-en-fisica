## 1. Condensadores y Capacitancia

Un condensador es un dispositivo compuesto por dos conductores (armaduras) con cargas iguales y opuestas $+Q$ y $-Q$, separados por una distancia $d$. Cuando se aplica una diferencia de potencial $V$, las cargas se acumulan en la superficie de las armaduras, creando un campo eléctrico uniforme interior.

La **capacitancia** o capacidad $C$ es una constante geométrica que mide la relación entre la carga almacenada en una de las placas y la diferencia de potencial aplicada:
$$C = \frac{Q}{V}$$
La capacitancia se mide en **Faradios (F)** en el SI ($1\text{ F} = 1\text{ C/V}$). Dado que un faradio es una unidad extremadamente grande, en circuitos reales se utilizan submúltiplos: milifaradio ($1\text{ mF} = 10^{-3}\text{ F}$), microfaradio ($1\text{ }\mu\text{F} = 10^{-6}\text{ F}$), nanofaradio ($1\text{ nF} = 10^{-9}\text{ F}$) y picofaradio ($1\text{ pF} = 10^{-12}\text{ F}$).

### 1.1 Dependencia de Factores Geométricos
Para un condensador de placas paralelas infinitas de área $A$ separadas por una distancia en el vacío $d$:
1. **Proporcionalidad al área ($C \propto A$):** Placas con mayor área disponen de más superficie física para que las cargas se distribuyan, permitiendo almacenar más carga por voltio.
2. **Inversa a la separación ($C \propto 1/d$):** Al reducir la distancia $d$, el campo eléctrico se concentra en un volumen menor y aumenta la fuerza de atracción electrostática entre las cargas de las placas opuestas, facilitando su acumulación.

La fórmula de capacitancia para placas paralelas en el vacío es:
$$C_0 = \epsilon_0 \frac{A}{d}$$
Donde $\epsilon_0 \approx 8.854 \times 10^{-12}\text{ F/m}$ es la permitividad eléctrica del vacío. Si insertamos un material dieléctrico de constante dieléctrica $\kappa$, la capacitancia aumenta a:
$$C = \epsilon \frac{A}{d} = \kappa \epsilon_0 \frac{A}{d}$$


## 2. Energía Almacenada y Transitorio RC

### 2.1 Energía del Campo Eléctrico
La energía potencial electrostática acumulada en el campo eléctrico de un condensador se calcula como el trabajo realizado para cargar el dispositivo:
$$W_C = \frac{1}{2} C V^2 = \frac{1}{2} Q V = \frac{Q^2}{2C}$$

### 2.2 Análisis del Transitorio RC y Corrección de Errata Teórica
Consideremos el ejercicio planteado en el texto guía (páginas 5-6): un condensador de $C = 50\ \mu\text{F}$ cargado inicialmente a $V_0 = 200\text{ V}$ se descarga a través de una lámpara con resistencia $R_{descarga} = 100\ \Omega$ durante un tiempo de descarga $t_{descarga} = 0.01\text{ s}$.

> [!IMPORTANT]
> **Identificación de la Errata del Texto Base:**
> El texto guía afirma que la energía disipada es:
> $$W_{disipada} = \frac{1}{2} C (V_0 - V_{final})^2$$
> Y sostiene que como la lámpara tiene resistencia interna, $V_{final} = V_0$, dando como resultado $W_{disipada} = 0\text{ J}$.
>
> Esta formulación física contiene dos inconsistencias conceptuales:
> 1. **Diferencia no lineal:** La variación de energía del condensador es la diferencia de los cuadrados de la tensión, no el cuadrado de la diferencia:
>    $$\Delta W = W_{inicial} - W_{final} = \frac{1}{2} C (V_0^2 - V_{final}^2) \neq \frac{1}{2} C (V_0 - V_{final})^2$$
> 2. **Descarga exponencial real:** Un condensador en descarga no mantiene su tensión constante; esta cae exponencialmente según la ecuación transitoria:
>    $$V_C(t) = V_0 e^{-t / RC}$$
>    La constante de tiempo del transitorio de descarga es $\tau = R_{descarga} C = 100\ \Omega \times 50\ \mu\text{F} = 0.005\text{ s}$.
>    A los $t = 0.01\text{ s}$ (es decir, transcurridos exactamente $2\tau$), la tensión final del condensador es:
>    $$V_{final} = 200 e^{-0.01 / 0.005} = 200 e^{-2} \approx 27.067\text{ V}$$
>
> La energía disipada real en forma de calor y luz a través de la lámpara es:
> $$W_{disipada} = \frac{1}{2} C V_0^2 - \frac{1}{2} C V_{final}^2 = 10\text{ J} - 0.0183\text{ J} \approx 9.982\text{ J}$$
> Físicamente, casi toda la energía (aproximadamente el 99.82%) se disipa en la lámpara tras $2\tau$.


## 3. Clasificación de Condensadores y Baterías

A continuación se presenta una recopilación formal de las tecnologías descritas en el PDF:

### 3.1 Tipos de Condensadores

* **Cerámicos:** Muy compactos y de bajo costo. Poseen bajas inductancias y resistencias parásitas, lo que los hace ideales para filtrado de alta frecuencia (RF) y desacoplo (decoupling) en circuitos integrados (PCBs).
* **Electrolíticos (Aluminio y Tantalio):** Ofrecen altas capacidades en espacios reducidos. Utilizan un electrolito líquido o gel. Son polarizados y se usan para filtrado de rizado en fuentes de alimentación.
* **De película (Film):** Fabricados con poliéster, polipropileno o PET. Altamente estables en temperatura y frecuencia, con baja tolerancia y bajas pérdidas, ideales para temporización precisa y audio de alta fidelidad.
* **Variables (Trimmers):** Permiten ajustar de forma mecánica la separación o el área efectiva de las placas. Usados para sintonización de radiofrecuencia.
* **De mica:** Altamente precisos y estables, con bajísimas pérdidas. Usados en osciladores de alta frecuencia y equipos médicos.
* **Supercondensadores:** Basados en almacenamiento electrostático de doble capa (EDLC). Tienen capacidades extremadamente grandes (cercanas a las de baterías) con tiempos de carga/descarga rápidos. Usados en frenado regenerativo.

### 3.2 Tipos de Baterías (Almacenamiento Químico)

A diferencia de los condensadores (almacenamiento en campo eléctrico), las baterías almacenan energía en forma de energía química mediante reacciones redox:

* **Plomo-Ácido:** Tecnología clásica y robusta de bajo costo, capaz de entregar corrientes elevadas. Usadas en el arranque de automóviles.
* **Iones de Litio (Li-ion):** Alta densidad energética, livianas y de alta eficiencia. Usadas en dispositivos electrónicos portátiles y vehículos eléctricos.
* **Níquel-Cadmio (NiCd) y Níquel-Metal Hidruro (NiMH):** NiCd posee altas tasas de descarga pero es altamente contaminante. NiMH es una alternativa ecológica con buena densidad energética.
* **Sodio-Ión (Na-ion):** Alternativa en desarrollo que utiliza sodio (abundante y económico), proyectada para sustituir al litio en almacenamiento de red.
* **Baterías de Flujo:** Utilizan electrolitos líquidos almacenados en tanques externos. Permiten un diseño modular e independiente de potencia y capacidad. Usadas para almacenamiento de energía a gran escala en redes eléctricas.


## 4. Circuitos Eléctricos y Leyes de Kirchhoff

Un circuito eléctrico es una red interconectada de componentes activos y pasivos. Las leyes fundamentales que rigen su comportamiento son:

* **Ley de Ohm:** Establece la relación lineal entre tensión, corriente y resistencia para un conductor óhmico:
  $$V = I R$$
* **Ley de Corrientes de Kirchhoff (LCK):** Basada en la conservación de la carga. La suma de corrientes que entran a un nodo es igual a la suma de corrientes que salen:
  $$\sum I_{entrantes} = \sum I_{salientes}$$
* **Ley de Voltajes de Kirchhoff (LVK):** Basada en la conservación de la energía en un campo conservativo. La suma algebraica de las diferencias de potencial en cualquier lazo o malla cerrada es cero:
  $$\sum_{lazo} V_i = 0$$

### 4.1 Combinaciones de Circuitos
1. **En Serie:** La corriente $I$ es común para todos los componentes. El voltaje total es la suma de los voltajes parciales. La resistencia equivalente es la suma de las resistencias:
   $$R_{eq} = \sum R_i$$
2. **En Paralelo:** El voltaje $V$ es común para todos los componentes. La corriente total es la suma de las corrientes de rama. Las conductancias ($G = 1/R$) se suman:
   $$G_{eq} = \sum G_i \implies \frac{1}{R_{eq}} = \sum \frac{1}{R_i}$$


## 5. Dieléctricos en Condensadores

Un dieléctrico es un material no conductor (aislante) que se inserta entre las placas de un condensador. Al someterse a un campo eléctrico externo $\vec{E}_0$, las moléculas del material se polarizan (las cargas se separan ligeramente), alineando sus dipolos microscópicos.

Esto genera un campo eléctrico interno inducido $\vec{E}_{ind}$ que apunta en sentido opuesto al campo aplicado. El campo neto resultante $\vec{E}$ en el interior disminuye:
$$\vec{E} = \vec{E}_0 - \vec{E}_{ind} = \frac{\vec{E}_0}{\kappa}$$
Donde $\kappa \ge 1$ es la **constante dieléctrica** del material. Al reducir el campo para una carga constante, el potencial disminuye ($V = V_0 / \kappa$), lo que aumenta la capacitancia ($C = \kappa C_0$).

### 5.1 Propiedades Clave de los Dieléctricos
* **Rigidez Dieléctrica:** La intensidad máxima de campo eléctrico que el material puede soportar sin sufrir una ruptura eléctrica (ionización y conducción destructiva). Se mide en V/m.
* **Pérdida Dieléctrica:** Disipación de energía en forma de calor por la fricción interna de los dipolos al oscilar en campos alternos de alta frecuencia.
* **Coeficiente de Temperatura de la Capacitancia:** Describe el cambio relativo del valor de capacitancia por grado de temperatura.

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar estilos de graficación
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: CURVAS DE CARGA Y DESCARGA RC (EJERCICIO DEL LIBRO CORREGIDO)
# =======================================================================
C = 50e-6         # Capacidad: 50 uF
V0 = 200.0        # Voltaje inicial de carga: 200 V
R_carga = 500.0   # Resistencia de carga: 500 ohms
R_descarga = 100.0 # Resistencia de descarga (lámpara): 100 ohms

tau_c = R_carga * C       # Constante carga: 0.025 s
tau_d = R_descarga * C    # Constante descarga: 0.005 s

t_c = np.linspace(0, 5 * tau_c, 300)
t_d = np.linspace(0, 5 * tau_d, 300)

# Carga: V(t) = V0*(1 - e^(-t/tau_c))
V_c_vals = V0 * (1.0 - np.exp(-t_c / tau_c))
I_c_vals = (V0 / R_carga) * np.exp(-t_c / tau_c)

# Descarga: V(t) = V0 * e^(-t/tau_d)
V_d_vals = V0 * np.exp(-t_d / tau_d)
I_d_vals = -(V0 / R_descarga) * np.exp(-t_d / tau_d)

# Punto de evaluación a t = 0.01 s de descarga
t_corte = 0.01
V_corte = V0 * np.exp(-t_corte / tau_d)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Panel de Carga
ax1.plot(t_c, V_c_vals, 'r-', lw=2.5, label='Voltaje $V_C(t)$')
ax1.plot(t_c, I_c_vals * R_carga, 'b--', lw=1.8, label=r'Corriente $I(t) \cdot R_{carga}$')
ax1.set_title(r"Curvas Transitorias de Carga RC ($R = 500\ \Omega$)", fontweight='bold')
ax1.set_xlabel("Tiempo t (s)")
ax1.set_ylabel("Voltaje (V)")
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Panel de Descarga
ax2.plot(t_d, V_d_vals, 'crimson', lw=2.5, label='Voltaje $V_C(t)$')
ax2.plot(t_d, I_d_vals * R_descarga, 'navy', lw=1.8, ls='--', label=r'Corriente $I(t) \cdot R_{lamp}$')
ax2.scatter(t_corte, V_corte, color='black', s=85, zorder=5, 
            label=rf'$V_C({t_corte}\text{{ s}}) = {V_corte:.2f}\text{{ V}}$')
ax2.set_title(r"Curvas Transitorias de Descarga RC ($R_{lamp} = 100\ \Omega$)", fontweight='bold')
ax2.set_xlabel("Tiempo t (s)")
ax2.set_ylabel("Voltaje (V)")
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: RESOLUTOR MATRICIAL DE REDES DE KIRCHHOFF
# =======================================================================
# Ecuaciones lineales de lazo (KVL):
# Mesh A:  30*Ia - 20*Ib = 12
# Mesh B: -20*Ia + 35*Ib = -6
R_mat = np.array([
    [30.0, -20.0],
    [-20.0, 35.0]
])

V_vec = np.array([12.0, -6.0])

# Resolver el sistema mediante álgebra lineal
I_mesh = np.linalg.solve(R_mat, V_vec)
Ia, Ib = I_mesh[0], I_mesh[1]

# Corrientes de rama correspondientes
I1 = Ia
I3 = Ib
I2 = Ia - Ib

# Validación de LCK en el nodo central (I1 - I2 - I3 = 0)
lck_val = I1 - I2 - I3
assert np.allclose(lck_val, 0.0, atol=1e-12), "¡La validación de Kirchhoff falló!"

# Graficar
fig, ax = plt.subplots(figsize=(7.5, 4.5))
ramas = ['Rama 1 ($I_1$)', 'Rama 2 ($I_2$)', 'Rama 3 ($I_3$)']
corrientes = [I1, I2, I3]
colores = ['royalblue', 'forestgreen', 'crimson']

ax.bar(ramas, corrientes, color=colores, edgecolor='black', width=0.45, zorder=3)
ax.axhline(0, color='black', lw=0.9)

for i, val in enumerate(corrientes):
    ax.text(i, val + (0.015 if val >= 0 else -0.035), f"{val:.4f} A", ha='center', va='bottom', fontweight='bold')

ax.set_title("Corrientes de Rama en Circuito de Mallas Mixto", fontsize=12, fontweight='bold')
ax.set_ylabel("Intensidad de Corriente (A)")
ax.grid(True, ls='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 3: PERFIL DE CAMPO ELÉCTRICO Y POTENCIAL CON DIELÉCTRICO
# =======================================================================
d_sep = 1.0        # Distancia entre placas (m)
V_placas = 100.0   # Tensión total (V)
x_inicio = 0.3     # Inicio del bloque dieléctrico (m)
x_fin = 0.7        # Fin del bloque dieléctrico (m)
kappa_val = 4.0    # Constante dieléctrica

x_grid = np.linspace(0.0, d_sep, 500)
dx = x_grid[1] - x_grid[0]

# Calcular campos para mantener V constante:
# E_vacio * L_vacio + E_diel * L_diel = V_placas, donde E_diel = E_vacio / kappa
L_v = x_inicio + (d_sep - x_fin)   # 0.6 m
L_d = x_fin - x_inicio             # 0.4 m

E_vac = V_placas / (L_v + L_d / kappa_val)
E_die = E_vac / kappa_val

# Construir perfiles espaciales
E_prof = np.zeros_like(x_grid)
for i, val_x in enumerate(x_grid):
    if val_x < x_inicio or val_x > x_fin:
        E_prof[i] = E_vac
    else:
        E_prof[i] = E_die

# Potencial por integración del campo (V(x) = integral E(x') dx')
V_prof = np.zeros_like(x_grid)
for i in range(1, len(x_grid)):
    V_prof[i] = V_prof[i-1] + 0.5 * (E_prof[i-1] + E_prof[i]) * dx

# Graficar
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 7.5))

# Gráfico de Campo Eléctrico
ax1.plot(x_grid, E_prof, 'r-', lw=2.5, label='Campo Eléctrico $E(x)$')
ax1.axvspan(x_inicio, x_fin, color='orange', alpha=0.18, label=rf'Dieléctrico ($\kappa = {kappa_val}$)')
ax1.set_title("Perfil de Campo Eléctrico entre Placas con Dieléctrico", fontweight='bold')
ax1.set_xlabel("Posición x (m)")
ax1.set_ylabel("Campo Eléctrico E (V/m)")
ax1.set_ylim(0, 1.2 * E_vac)
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True, loc='upper right')

# Gráfico de Potencial
ax2.plot(x_grid, V_prof, 'b-', lw=2.5, label='Potencial $V(x)$')
ax2.axvspan(x_inicio, x_fin, color='orange', alpha=0.18, label=rf'Dieléctrico ($\kappa = {kappa_val}$)')
ax2.set_title("Perfil de Potencial Eléctrico $V(x)$", fontweight='bold')
ax2.set_xlabel("Posición x (m)")
ax2.set_ylabel("Potencial V (V)")
ax2.set_ylim(0, 1.1 * V_placas)
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True, loc='upper left')

plt.tight_layout()
plt.show()
```

## 7. Verificación Analítica con SymPy

Para validar las ecuaciones transitorias deducidas para el circuito RC de carga, resolveremos de forma simbólica la ecuación diferencial ordinaria lineal de primer orden de la carga en el condensador:
$$R \frac{dq}{dt} + \frac{q}{C} = V_0$$
Sujeta a la condición inicial de condensador descargado al inicio ($q(0) = 0$).


```python
import sympy as sp

# Definir variables y constantes físicas simbólicas reales positivas
t_sym = sp.symbols('t', real=True, positive=True)
R_sym, C_sym, V0_sym = sp.symbols('R C V_0', positive=True)
q_func = sp.Function('q')(t_sym)

# Definir la EDO del circuito RC
edo_rc = R_sym * sp.diff(q_func, t_sym) + q_func / C_sym - V0_sym

# Resolver simbólicamente la EDO con condición inicial q(0) = 0
sol_rc = sp.dsolve(edo_rc, ics={q_func.subs(t_sym, 0): 0})
sol_rc_simp = sp.simplify(sol_rc)

# Expresión analítica teórica manual para verificar
sol_teorica_rc = C_sym * V0_sym * (1.0 - sp.exp(-t_sym / (R_sym * C_sym)))
diferencia_rc = sp.simplify(sol_rc_simp.rhs - sol_teorica_rc)

print("======================================================================")
print("RESOLUCIÓN SIMBÓLICA DEL TRANSITORIO RC CON SYMPY")
print("======================================================================")
print(f"Ecuación Diferencial: {R_sym} * dq/dt + q/{C_sym} - {V0_sym} = 0")
print(f"Solución Calculada:   q(t) = {sol_rc_simp.rhs}")
print(f"Fórmula Esperada:     q(t) = {sol_teorica_rc}")
print(f"Diferencia simbólica: {diferencia_rc} (Esperado: 0)")

assert diferencia_rc == 0, "Error: ¡Las soluciones del transitorio RC no coinciden!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! La solución transitoria se validó con éxito.")
print("======================================================================")
```

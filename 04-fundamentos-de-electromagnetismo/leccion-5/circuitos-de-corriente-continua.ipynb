## 1. Corriente Eléctrica Continua y Fuentes

La corriente eléctrica se define como la tasa neta a la que fluye la carga a través de una sección transversal de un conductor:
$$I = \frac{dq}{dt}$$
La unidad en el SI es el **Amperio (A)** ($1\text{ A} = 1\text{ C/s}$).

La **corriente continua (CC / CD)** es aquella en la que los portadores de carga se desplazan de forma unidireccional y su magnitud y sentido permanecen constantes en el tiempo. Por convención histórica (debida a Benjamin Franklin), el sentido de la corriente convencional se define desde el polo positivo (+) al polo negativo (-), el cual es opuesto al desplazamiento físico real de los electrones libres (que viajan del polo negativo al positivo).

### 1.1 Análisis y Corrección de Erratas del Libro Guía

> [!IMPORTANT]
> **Errata 1: Clasificación de Baterías (pág 4)**
> El texto guía coloca las baterías de Níquel-Cadmio (NiCd) dentro del grupo de baterías primarias (no recargables). 
> * **Corrección Física:** Las baterías de NiCd son baterías secundarias (recargables). Almacenan energía mediante reacciones redox reversibles de hidróxido de óxido de níquel y cadmio metálico, permitiendo cientos de ciclos de recarga, aunque actualmente están en desuso comercial por la toxicidad del cadmio.

> [!IMPORTANT]
> **Errata 2: Eficiencia de Rectificadores (pág 4)**
> El texto guía indica que los rectificadores de media onda poseen una eficiencia del 50% y los de onda completa del 100%.
> * **Corrección Física:** La eficiencia de conversión de potencia de corriente continua (DC power efficiency, $\eta$) de un rectificador ideal alimentado por una onda sinusoidal pura se define como $\eta = P_{DC} / P_{AC}$. Para diodos ideales:
>   * Rectificador de media onda: $\eta_{max} = \frac{4}{\pi^2} \approx 40.53\%$
>   * Rectificador de onda completa: $\eta_{max} = \frac{8}{\pi^2} \approx 81.06\%$
>   Los valores del libro (50% y 100%) representan una idealización simplificada que confunde la fracción del ciclo activo con la eficiencia real de potencia disipada.


## 2. Resistencia Eléctrica y Reducción de Redes

La resistencia $R$ de un elemento conductor óhmico uniforme depende de su geometría y del material (resistividad $\rho$):
$$R = \rho \frac{l}{A}$$
Donde $l$ es la longitud y $A$ la sección transversal. La **Ley de Ohm** relaciona la tensión y corriente como:
$$V = I R$$

### 2.1 Reducción de Resistencias
* **Conexión en Serie:** La corriente es la misma en todas las resistencias, los voltajes se suman. La resistencia equivalente es:
  $$R_{eq} = \sum_{n=1}^{N} R_n$$
* **Conexión en Paralelo:** El voltaje es el mismo en todas las resistencias, las corrientes se dividen. La resistencia equivalente es:
  $$\frac{1}{R_{eq}} = \sum_{n=1}^{N} \frac{1}{R_n} \implies R_{eq} = \frac{R_1 R_2}{R_1 + R_2} \quad (\text{para } N=2)$$

### 2.2 Resolución del Ejemplo 5 (pág 11-12)
Dada una red mixta compleja con terminales de entrada/salida y las resistencias: $R_1 = 4\ \Omega$, $R_2 = 2\ \Omega$, $R_3 = 1\ \Omega$, $R_4 = 5\ \Omega$, $R_5 = 3\ \Omega$, $R_6 = 6\ \Omega$ y $R_7 = 8\ \Omega$:
1. **Paso 1:** $R_5$ y $R_6$ están en paralelo:
   $$R_{56} = \frac{R_5 R_6}{R_5 + R_6} = \frac{3 \cdot 6}{3 + 6} = 2\ \Omega$$
2. **Paso 2:** $R_3$ y $R_4$ están en serie:
   $$R_{34} = R_3 + R_4 = 1 + 5 = 6\ \Omega$$
3. **Paso 3:** $R_2$ y $R_{56}$ están en serie:
   $$R_{256} = R_2 + R_{56} = 2 + 2 = 4\ \Omega$$
4. **Paso 4:** $R_{256}$ y $R_{34}$ están en paralelo:
   $$R_{eq1} = \frac{R_{256} R_{34}}{R_{256} + R_{34}} = \frac{4 \cdot 6}{4 + 6} = 2.4\ \Omega$$
5. **Paso 5:** La resistencia equivalente total entre terminales (con $R_1$, $R_{eq1}$ y $R_7$ en serie) es:
   $$R_{eq} = R_1 + R_{eq1} + R_7 = 4 + 2.4 + 8 = 14.4\ \Omega$$


## 3. Leyes de Kirchhoff

Para redes complejas que no pueden reducirse a combinaciones serie-paralelo directas, aplicamos las **Leyes de Kirchhoff**:

1. **Ley de Corrientes (LCK - Ley de Nodos):** En cualquier nodo de la red, la suma algebraica de las corrientes es cero:
   $$\sum_{n=1}^{N} I_n = 0$$
2. **Ley de Voltajes (LVK - Ley de Mallas):** La suma de las diferencias de potencial eléctrico alrededor de cualquier malla o lazo cerrado es cero:
   $$\sum_{m=1}^{M} V_m = 0$$

### 3.1 Resolución del Ejemplo 9 (pág 16-17)
Consideremos un circuito de dos mallas con una batería de $10\text{ V}$ a la izquierda, una resistencia superior izquierda de $2\ \Omega$, una rama central con una resistencia de $8\ \Omega$, una resistencia superior derecha de $4\ \Omega$ y una batería a la derecha de $6\text{ V}$ (con polaridad invertida, positivo al fondo y negativo al tope, por lo que actúa en serie aditiva en la malla):

Planteamos el sistema con corrientes de rama: $I_1$ (izquierda), $I_2$ (centro, hacia abajo) e $I_3$ (derecha, hacia la derecha):
* **Nodo central superior (LCK):** 
  $$I_1 - I_2 - I_3 = 0 \implies I_1 = I_2 + I_3$$
* **Malla A (LVK - izquierda, sentido horario):**
  $$10 - 2 I_1 - 8 I_2 = 0 \implies 2 I_1 + 8 I_2 = 10$$
* **Malla B (LVK - derecha, sentido antihorario):**
  $$-6 + 4 I_3 - 8 I_2 = 0 \implies -8 I_2 + 4 I_3 = 6$$

Resolviendo el sistema:
1. Despejando de Malla A: $I_1 = 5 - 4 I_2$
2. Despejando de Malla B: $I_3 = 1.5 + 2 I_2$
3. Sustituyendo en la LCK:
   $$5 - 4 I_2 = I_2 + (1.5 + 2 I_2) \implies 5 - 1.5 = 7 I_2 \implies 3.5 = 7 I_2 \implies I_2 = 0.5\text{ A}$$
4. Corrientes resultantes:
   $$I_1 = 3\text{ A}, \quad I_2 = 0.5\text{ A}, \quad I_3 = 2.5\text{ A}$$
5. Caídas de voltaje:
   $$V_{R1} = 6\text{ V}, \quad V_{R2} = 4\text{ V}, \quad V_{R3} = 10\text{ V}$$


## 4. Circuitos RC y Respuesta Transitoria

Un circuito RC contiene una combinación en serie de una resistencia $R$ y un condensador $C$. A diferencia de los resistores puros, la corriente y la carga en un circuito RC varían en el tiempo.

### 4.1 Proceso de Carga
Al conectar una fuente de voltaje $V_0$ a un circuito RC descargado en $t=0$, la carga $q(t)$, el voltaje $V_C(t)$ y la corriente $i(t)$ son:
$$q(t) = Q_{max} (1 - e^{-t/RC}) = C V_0 (1 - e^{-t/RC})$$
$$V_C(t) = V_0 (1 - e^{-t/RC})$$
$$i(t) = I_{max} e^{-t/RC} = \frac{V_0}{R} e^{-t/RC}$$

### 4.2 Proceso de Descarga y Corrección de Errata de Corriente
Al desconectar la fuente y cortocircuitar el capacitor cargado inicialmente con $Q_0$ a través de $R$ en $t=0$:
$$q(t) = Q_0 e^{-t/RC}$$
$$V_C(t) = V_0 e^{-t/RC}$$

> [!IMPORTANT]
> **Corrección de Errata de Corriente de Descarga (pág 18):**
> El texto base repite la fórmula de carga para la corriente o escribe de forma errónea $i = Q e^{-t/RC}$ (donde mezcla unidades de carga y corriente).
> * **Fórmula Física Correcta:** La corriente es la derivada temporal de la carga: $i(t) = \frac{dq}{dt} = -\frac{Q_0}{RC} e^{-t/RC} = -I_0 e^{-t/RC}$. El signo negativo representa físicamente que la corriente fluye en sentido opuesto al de carga durante el drenado del capacitor.

### 4.3 Constante de Tiempo ($\tau$)
El producto $\tau = R C$ se denomina **constante de tiempo** del circuito y mide la rapidez de respuesta del transitorio. Representa el tiempo requerido para que:
* En la carga: el condensador alcance el 63.2% de su carga máxima.
* En la descarga: la carga disminuya hasta el 36.8% de su valor inicial.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar estilos estéticos
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: REDUCTOR AUTOMÁTICO DE RESISTENCIAS MIXTAS
# =======================================================================
R1 = 4.0
R2 = 2.0
R3 = 1.0
R4 = 5.0
R5 = 3.0
R6 = 6.0
R7 = 8.0

print("--- Reducción Paso a Paso del Ejemplo 5 ---")
R56 = (R5 * R6) / (R5 + R6)
print(f"Paso 1 (R5 || R6): {R5} || {R6} = {R56:.2f} ohms")

R34 = R3 + R4
print(f"Paso 2 (R3 + R4): {R3} + {R4} = {R34:.2f} ohms")

R256 = R2 + R56
print(f"Paso 3 (R2 + R56): {R2} + {R56:.2f} = {R256:.2f} ohms")

Req1 = (R256 * R34) / (R256 + R34)
print(f"Paso 4 (R256 || R34): {R256:.2f} || {R34:.2f} = {Req1:.2f} ohms")

Req_total = R1 + Req1 + R7
print(f"Paso 5 (Req Total): {R1} + {Req1:.2f} + {R7} = {Req_total:.2f} ohms")

# Graficar la evolución de la reducción de la resistencia total ficticia
pasos = ['Inicial', 'R5||R6', 'R3+R4', 'R2+R56', 'R256||R34', 'Req Total']
valores = [
    R1 + R2 + R3 + R4 + R5 + R6 + R7,
    R1 + R2 + R3 + R4 + R56 + R7,
    R1 + R2 + R34 + R56 + R7,
    R1 + R256 + R34 + R7,
    R1 + Req1 + R7,
    Req_total
]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(pasos, valores, marker='o', color='darkgreen', lw=2.5, markersize=8)
for i, val in enumerate(valores):
    ax.text(i, val + 1.0, rf"{val:.2f} $\Omega$", ha='center', fontweight='bold')

ax.set_title("Evolución de la Resistencia Equivalente durante la Reducción", fontweight='bold')
ax.set_xlabel("Etapa de Simplificación")
ax.set_ylabel(r"Suma de Resistencias ($\Omega$)")
ax.grid(True, ls='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: RESOLUTOR DE MALLAS Y BALANCE DE POTENCIA (EJEMPLO 9)
# =======================================================================
# Planteamiento matricial del Ejemplo 9:
#  2*I1 + 8*I2 + 0*I3 = 10
#  0*I1 - 8*I2 + 4*I3 = 6
#  1*I1 - 1*I2 - 1*I3 = 0
A_mat = np.array([
    [2.0, 8.0, 0.0],
    [0.0, -8.0, 4.0],
    [1.0, -1.0, -1.0]
])

B_vec = np.array([10.0, 6.0, 0.0])

I_sol = np.linalg.solve(A_mat, B_vec)
I1_val, I2_val, I3_val = I_sol[0], I_sol[1], I_sol[2]

# Calcular potencias asociadas a cada componente
P_del_10V = 10.0 * I1_val
P_del_6V = 6.0 * I3_val
P_del_total = P_del_10V + P_del_6V

P_dis_R1 = (I1_val**2) * 2.0
P_dis_R2 = (I2_val**2) * 8.0
P_dis_R3 = (I3_val**2) * 4.0
P_dis_total = P_dis_R1 + P_dis_R2 + P_dis_R3

print("--- Resultados de Corrientes de Rama ---")
print(f"I1 (Rama Izquierda): {I1_val:.3f} A")
print(f"I2 (Rama Central):   {I2_val:.3f} A")
print(f"I3 (Rama Derecha):   {I3_val:.3f} A")

# Comprobar conservación de energía
assert np.isclose(P_del_total, P_dis_total), "¡El balance de energía falló!"

# Graficar balance de potencia
fig, ax = plt.subplots(figsize=(7.5, 4.5))
elementos = ['Batería 10V', 'Batería 6V', r'R1 (2 $\Omega$)', r'R2 (8 $\Omega$)', r'R3 (4 $\Omega$)']
potencias = [P_del_10V, P_del_6V, -P_dis_R1, -P_dis_R2, -P_dis_R3]
colores = ['royalblue', 'deepskyblue', 'crimson', 'firebrick', 'darkred']

ax.bar(elementos, potencias, color=colores, edgecolor='black', width=0.45, zorder=3)
ax.axhline(0, color='black', lw=0.9)
for i, val in enumerate(potencias):
    ax.text(i, val + (0.8 if val >= 0 else -1.8), f"{abs(val):.2f} W", ha='center', va='bottom', fontweight='bold')

ax.set_title("Balance de Potencia en el Circuito de Dos Mallas", fontsize=12, fontweight='bold')
ax.set_ylabel("Potencia (W) [Positiva = Entregada, Negativa = Disipada]")
ax.grid(True, ls='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 3: CURVAS DINÁMICAS DE TRANSITORIO RC (EJEMPLO 10)
# =======================================================================
V_fuente = 12.0     # Voltaje de fuente: 12 V
C_cap = 5e-6        # Capacidad: 5 uF
R_res = 8e5         # Resistencia: 800 kOhms

tau_val = R_res * C_cap     # Constante: 4.0 s
Q_max_val = C_cap * V_fuente # Carga max: 60 uC
I_max_val = V_fuente / R_res # Corriente max: 15 uA

t_grid = np.linspace(0, 5 * tau_val, 300)

V_C_t = V_fuente * (1.0 - np.exp(-t_grid / tau_val))
q_t_vals = Q_max_val * (1.0 - np.exp(-t_grid / tau_val)) * 1e6 # uC
i_t_vals = I_max_val * np.exp(-t_grid / tau_val) * 1e6        # uA

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Panel de Voltaje y Carga
ax1.plot(t_grid, V_C_t, 'darkorange', lw=2.5, label='Voltaje $V_C(t)$ (V)')
ax1.plot(t_grid, q_t_vals / 5.0, 'navy', lw=1.8, ls='--', label=r'Carga $q(t)/5$ (\mu C)')
for k in range(1, 6):
    ax1.axvline(k * tau_val, color='gray', ls=':', alpha=0.45)
    if k == 1:
        ax1.text(tau_val + 0.15, 2.5, r'$1\tau = 4\text{ s}$', fontsize=9, color='gray')
ax1.set_title("Voltaje y Carga durante la Carga RC", fontweight='bold')
ax1.set_xlabel("Tiempo t (s)")
ax1.set_ylabel("Voltaje (V) / Escala de Carga")
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Panel de Corriente
ax2.plot(t_grid, i_t_vals, 'forestgreen', lw=2.5, label=r'Corriente $I(t)$ (\mu A)')
for k in range(1, 6):
    ax2.axvline(k * tau_val, color='gray', ls=':', alpha=0.45)
    if k == 1:
        ax2.text(tau_val + 0.15, 10.0, r'$1\tau = 4\text{ s}$', fontsize=9, color='gray')
ax2.set_title("Corriente durante la Carga RC", fontweight='bold')
ax2.set_xlabel("Tiempo t (s)")
ax2.set_ylabel(r"Corriente (\mu A)")
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

## 6. Verificación Analítica con SymPy

Utilizaremos SymPy para resolver simbólicamente el sistema lineal de ecuaciones algebraicas del circuito de dos mallas del Ejemplo 9, validando la exactitud de las corrientes resueltas numéricamente.


```python
import sympy as sp

# Definir corrientes como variables simbólicas reales
I1_sym, I2_sym, I3_sym = sp.symbols('I_1 I_2 I_3', real=True)

# Definir ecuaciones lineales correspondientes al Ejemplo 9
eq1 = sp.Eq(2 * I1_sym + 8 * I2_sym, 10)       # Malla A
eq2 = sp.Eq(-8 * I2_sym + 4 * I3_sym, 6)       # Malla B
eq3 = sp.Eq(I1_sym - I2_sym - I3_sym, 0)       # Nodo

# Resolver el sistema lineal simbólicamente
solucion_simb = sp.solve([eq1, eq2, eq3], (I1_sym, I2_sym, I3_sym))

print("======================================================================")
print("RESOLUCIÓN SIMBÓLICA DE LEYES DE KIRCHHOFF CON SYMPY")
print("======================================================================")
print(f"Solución Simbólica: {solucion_simb}")
print(f"I1 (Esperado: 3 A):  {solucion_simb[I1_sym]} A")
print(f"I2 (Esperado: 0.5 A):{solucion_simb[I2_sym]} A")
print(f"I3 (Esperado: 2.5 A):{solucion_simb[I3_sym]} A")

assert solucion_simb[I1_sym] == 3 and solucion_simb[I2_sym] == sp.Rational(1, 2) and solucion_simb[I3_sym] == sp.Rational(5, 2), \
    "Error: ¡La resolución simbólica de Kirchhoff no coincide con el libro!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Se validaron de forma exacta las leyes de mallas.")
print("======================================================================")
```

## 1. Flujo Magnético

El **flujo magnético** $\Phi_B$ a través de una superficie abierta mide la cantidad neta de líneas de campo magnético que atraviesan dicha área. Matemáticamente se define como la integral del producto escalar del campo $\vec{B}$ y el elemento diferencial de área $d\vec{A}$:
$$\Phi_B = \int \vec{B} \cdot d\vec{A}$$

Si el campo magnético $\vec{B}$ es uniforme en una superficie plana de área total $A$, la integral se simplifica a:
$$\Phi_B = B A \cos\theta$$
Donde $\theta$ es el ángulo entre la dirección del campo magnético y el vector normal perpendicular a la superficie.

* **Unidades:** En el SI el flujo magnético se mide en **Webers (Wb)**:
  $$1\text{ Wb} = 1\text{ T} \cdot \text{m}^2$$
* **Ley de Gauss para el Magnetismo:** Si la superficie es cerrada, el flujo neto siempre es cero ($\oint \vec{B} \cdot d\vec{A} = 0$), lo que postula la inexistencia de monopolos magnéticos aislados.

### 1.1 Aplicación: Ejemplo 1 de Superficie Inclinada
Consideremos una espira plana de área $A = 3\text{ cm}^2 = 3 \times 10^{-4}\text{ m}^2$ expuesta a un campo magnético uniforme. Si el flujo es de $\Phi_B = 0.9\text{ mWb} = 9 \times 10^{-4}\text{ Wb}$ y el vector normal de la espira forma un ángulo de $\theta = 60^\circ$ con el campo, determinamos la magnitud del campo magnético:
$$B = \frac{\Phi_B}{A \cos\theta} = \frac{9 \times 10^{-4}\text{ Wb}}{3 \times 10^{-4}\text{ m}^2 \cdot \cos(60^\circ)} = \frac{9 \times 10^{-4}}{3 \times 10^{-4} \cdot 0.5} = 6.0\text{ T}$$


## 2. Inducción de Fem y Ley de Faraday

La **inducción electromagnética** se descubrió de forma independiente por Michael Faraday en Inglaterra y Joseph Henry en Estados Unidos en 1831. Descubrieron que un campo magnético variable en el tiempo es capaz de generar una corriente eléctrica en una espira conductora.

La **Ley de Faraday** establece que la magnitud de la fuerza electromotriz (fem) $\mathcal{E}$ inducida en una trayectoria cerrada es igual a la tasa de variación temporal del flujo magnético a través del área encerrada por la trayectoria:
$$\mathcal{E} = -\frac{d\Phi_B}{dt}$$
Si la bobina contiene $N$ espiras idénticas acopladas, la fem inducida neta es:
$$\mathcal{E} = -N \frac{d\Phi_B}{dt}$$

Dado que $\Phi_B = B A \cos\theta$, se puede inducir una fem si varía en el tiempo cualquiera de estas tres magnitudes (o una combinación de ellas):
1. La magnitud del campo magnético $B(t)$.
2. El área de la espira expuesta al campo $A(t)$.
3. El ángulo de orientación $\theta(t)$ de la espira respecto al campo.

### 2.1 Análisis y Corrección de Errata de Transposición (Ejemplo 2, pág 10)

> [!IMPORTANT]
> **Errata Numérica en el Ejemplo 2:**
> El Ejemplo 2 del libro describe una bobina circular de $N=500$ espiras de radio $R=4\text{ cm} = 0.04\text{ m}$ (área $A = \pi R^2 \approx 0.0050265\text{ m}^2$) en un campo uniforme inclinado $60^\circ$ con respecto al plano de la bobina (lo que significa que el ángulo con la normal de la espira es $\theta = 30^\circ$). El campo disminuye a razón de $dB/dt = -0.2\text{ T/s}$. El libro escribe como resultado final:
> $$\mathcal{E} = 0.453\text{ V}$$
> * **Corrección Física:** Esto es una errata de transposición tipográfica al digitar la mantisa del resultado. Calculando de forma exacta con la ley de Faraday:
>   $$\mathcal{E} = -N \frac{d(B A \cos\theta)}{dt} = -N A \cos\theta \frac{dB}{dt} = -(500) \cdot (0.0050265\text{ m}^2) \cdot \cos(30^\circ) \cdot (-0.2\text{ T/s})$$
>   $$\mathcal{E} = 100 \cdot (0.0050265) \cdot 0.866025 \approx 0.4353\text{ V} \approx 0.436\text{ V}$$
>   El libro guía intercambió los dígitos $3$ y $5$ de la mantisa escribiendo $0.453\text{ V}$ en lugar de $0.436\text{ V}$ (una diferencia de aproximadamente el $4\%$). Corregiremos este valor en el desarrollo del cuaderno.


## 3. Ley de Lenz

El signo negativo en la Ley de Faraday representa la dirección de la fem inducida. Este sentido físico se rige por la **Ley de Lenz**:

> **Enunciado:** La corriente inducida en una espira cerrada fluye en una dirección tal que el campo magnético generado por dicha corriente se opone al cambio de flujo magnético original que la produjo.

### 3.1 Análisis de Casos
* **Flujo Creciente:** Si el flujo magnético original a través de la espira aumenta, la corriente inducida genera un campo magnético de sentido opuesto al campo aplicado para contrarrestar el incremento.
* **Flujo Decreciente:** Si el flujo original disminuye, la corriente inducida genera un campo en el mismo sentido que el campo aplicado para reforzarlo y contrarrestar la disminución.

Este comportamiento es una manifestación directa del principio de conservación de la energía: si la corriente inducida reforzara el cambio de flujo original, se produciría un bucle de retroalimentación infinita que generaría energía de la nada.


## 4. Fem en Movimiento y Campo Eléctrico Inducido

### 4.1 Fuerza Electromotriz de Movimiento
Cuando un conductor recto de longitud $l$ se desliza a velocidad constante $v$ en una trayectoria perpendicular a un campo magnético uniforme $\vec{B}$, los electrones libres dentro del conductor experimentan una fuerza magnética lateral de Lorentz $\vec{F}_B = q \vec{v} \times \vec{B}$. esto empuja a las cargas hacia un extremo, polarizando eléctricamente el conductor y generando una diferencia de potencial estable.

La magnitud de esta **fem de movimiento** viene dada por:
$$\mathcal{E} = v B l$$
Esta fórmula es equivalente a la ley de Faraday considerando que el desplazamiento del conductor móvil varía de forma lineal el área efectiva encerrada por el circuito en el tiempo ($dA/dt = l \cdot v$).

### 4.2 Aplicación: Ejemplo del Avión Comercial
Un avión comercial viaja a una velocidad constante de $1000\text{ km/h} \approx 277.78\text{ m/s}$ en una región donde el campo terrestre vertical es de $B = 5 \times 10^{-5}\text{ T}$. La envergadura de las alas es de $l = 70\text{ m}$. La diferencia de potencial inducida en los extremos de las alas es:
$$\mathcal{E} = v B l = (277.78\text{ m/s}) \cdot (5 \times 10^{-5}\text{ T}) \cdot (70\text{ m}) \approx 0.972\text{ V} \approx 1.0\text{ V}$$

### 4.3 Campo Eléctrico Inducido
La Ley de Faraday posee una implicación más profunda. Un campo magnético variable en el tiempo induce un campo eléctrico $\vec{E}$ en el espacio, incluso en ausencia de conductores físicos. La relación general es:
$$\oint \vec{E} \cdot d\vec{l} = -\frac{d\Phi_B}{dt}$$
A diferencia del campo eléctrico electrostático (producido por cargas en reposo, que es conservativo y tiene $\oint \vec{E} \cdot d\vec{l} = 0$), el campo eléctrico inducido **no es conservativo** y sus líneas de campo forman curvas cerradas continuas.


## 5. Autoinductancia e Inductancia Mutua

### 5.1 Autoinductancia (Inductancia $L$)
Cuando circula una corriente variable $i(t)$ por un circuito, el campo magnético generado por la corriente también varía en el tiempo. Esta variación induce una fem en el propio circuito que se opone al cambio de corriente. Este fenómeno se conoce como **autoinducción**.

La **autoinductancia** (o simplemente inductancia) $L$ de un componente (inductor) se define como la razón entre el flujo magnético encadenado y la corriente que lo produce:
$$L = \frac{N\Phi_B}{i}$$
La fem autoinducida viene dada por:
$$\mathcal{E}_L = -L \frac{di}{dt}$$
* **Unidades:** En el SI la inductancia se mide en **Henrys (H)**:
  $$1\text{ H} = 1\text{ }\frac{\text{V} \cdot \text{s}}{\text{A}} = 1\text{ }\Omega \cdot \text{s}$$
* La inductancia es una medida de la inercia eléctrica del circuito, es decir, su oposición al cambio abrupto de corriente.

### 5.2 Inductancia Mutua ($M$)
Si colocamos dos bobinas cercanas, una corriente variable en la bobina 1 induce una fem en la bobina 2 por flujo acoplado:
$$\mathcal{E}_2 = -M \frac{di_1}{dt} \quad \text{y} \quad \mathcal{E}_1 = -M \frac{di_2}{dt}$$
Donde la constante de inductancia mutua $M$ depende únicamente de la geometría y la orientación relativa de ambas bobinas:
$$M = \frac{N_2 \Phi_{B2}}{i_1} = \frac{N_1 \Phi_{B1}}{i_2}$$


## 6. Energía Magnética

Para establecer una corriente en un inductor, la fuente de energía externa debe realizar trabajo contra la fem autoinducida. Este trabajo se almacena como energía potencial en el campo magnético del inductor.

### 6.1 Energía Almacenada en un Inductor
La potencia eléctrica instantánea suministrada al inductor es $P = v_{ab} i = L i \frac{di}{dt}$. La energía infinitesimal entregada en un tiempo $dt$ es:
$$dU = P dt = L i di$$
Integrando desde corriente cero hasta un valor final $I$:
$$U = \int_{0}^{I} L i di \implies U = \frac{1}{2} L I^2$$

### 6.2 Aplicación: Ejemplo 4 de Almacenamiento Energético
Queremos almacenar una energía de $1\text{ kWh} = 3.6 \times 10^6\text{ J}$ en un inductor portador de una corriente de $I = 200\text{ A}$. Calculamos la inductancia requerida:
$$L = \frac{2 U}{I^2} = \frac{2 \cdot (3.6 \times 10^6\text{ J})}{(200\text{ A})^2} = \frac{7.2 \times 10^6}{40000} = 180\text{ H}$$

### 6.3 Densidad de Energía Magnética
La energía almacenada por unidad de volumen ($u_B$) en un campo magnético uniforme de magnitud $B$ en un medio con permeabilidad magnética $\mu$ es:
$$u_B = \frac{B^2}{2\mu}$$
En el vacío se escribe $u_B = \frac{B^2}{2\mu_0}$.


## 7. Circuitos RL

Un circuito que contiene un resistor de resistencia $R$ y un inductor de inductancia $L$ conectados en serie se denomina **circuito RL**.

### 7.1 Crecimiento de Corriente (Carga)
Si en un circuito RL inicialmente abierto con corriente cero cerramos el lazo en $t=0$ conectándolo a una batería de fem $\mathcal{E}$:
$$\mathcal{E} - i R - L \frac{di}{dt} = 0$$
Resolviendo esta ecuación diferencial ordinaria de primer orden lineal por separación de variables (con condición inicial $i(0) = 0$), obtenemos:
$$i(t) = \frac{\mathcal{E}}{R} \left(1 - e^{-t/\tau}\right)$$
Donde la **constante de tiempo inductiva** del circuito es:
$$\tau = \frac{L}{R}$$
En un tiempo $t = \tau$, la corriente alcanza aproximadamente el $63.2\%$ de su valor final límite $I_{máx} = \mathcal{E}/R$.

### 7.2 Decrecimiento de Corriente (Descarga)
Si la corriente en el circuito tiene un valor inicial estable $I_0$ y en $t=0$ desconectamos la batería cerrando el bucle con un puente conductor:
$$-i R - L \frac{di}{dt} = 0 \implies i(t) = I_0 e^{-t/\tau}$$
La corriente decrece exponencialmente hacia cero. En un tiempo $t = \tau$, la corriente ha decaído a aproximadamente el $36.8\%$ de su valor inicial.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: TRANSITORIOS EN CIRCUITOS RL (CARGA Y DESCARGA)
# =======================================================================
V_f = 12.0          # Tensión de fuente (V)
R_r = 8.0           # Resistencia (Ohms)
L_i = 4.0           # Inductancia (H)

tau_ind = L_i / R_r # Constante de tiempo inductiva: 0.5 s
I_max_val = V_f / R_r # Corriente límite: 1.5 A

t_array = np.linspace(0, 5 * tau_ind, 500)

# Comportamientos durante la carga (crecimiento)
i_carga = I_max_val * (1.0 - np.exp(-t_array / tau_ind))
v_R_carga = i_carga * R_r
v_L_carga = V_f * np.exp(-t_array / tau_ind)

# Comportamientos durante la descarga (decrecimiento)
i_descarga = I_max_val * np.exp(-t_array / tau_ind)
v_R_descarga = i_descarga * R_r
v_L_descarga = -V_f * np.exp(-t_array / tau_ind)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5.5))

# Panel 1: Crecimiento de corriente
ax1.plot(t_array, i_carga, 'forestgreen', lw=2.5, label='Corriente $i(t)$ (A)')
ax1.plot(t_array, v_L_carga / R_r, 'navy', lw=1.8, ls='--', label='Tensión Inductor $v_L(t)/R$')
for k in range(1, 6):
    ax1.axvline(k * tau_ind, color='gray', ls=':', alpha=0.45)
    if k == 1:
        ax1.text(tau_ind + 0.02, 0.4, r'$1\tau = 0.5\text{ s}$', color='gray', fontsize=9.5)
ax1.set_title("Crecimiento de Corriente en Circuito RL", fontweight='bold')
ax1.set_xlabel("Tiempo t (s)")
ax1.set_ylabel("Corriente (A) / Escala de Tensión")
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Panel 2: Decrecimiento de corriente
ax2.plot(t_array, i_descarga, 'crimson', lw=2.5, label='Corriente $i(t)$ (A)')
ax2.plot(t_array, v_L_descarga / R_r, 'darkred', lw=1.8, ls='--', label='Tensión Inductor $v_L(t)/R$')
for k in range(1, 6):
    ax2.axvline(k * tau_ind, color='gray', ls=':', alpha=0.45)
    if k == 1:
        ax2.text(tau_ind + 0.02, 0.4, r'$1\tau = 0.5\text{ s}$', color='gray', fontsize=9.5)
ax2.set_title("Decaimiento de Corriente en Circuito RL", fontweight='bold')
ax2.set_xlabel("Tiempo t (s)")
ax2.set_ylabel("Corriente (A) / Escala de Tensión")
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: ESPIRA MÓVIL EN CAMPO MAGNETICO - FLUJO, FEM Y LENZ
# =======================================================================
width = 0.1       # Ancho de la espira en X (m)
length = 0.1      # Longitud de la espira en Y (m)
R_loop = 2.0      # Resistencia de la espira (Ohms)
B_field_z = 1.5   # Campo magnético en Z (T) (en zona x > 0)
v_vel = 2.0       # Velocidad de traslación en X (m/s)

x_front = np.linspace(-0.15, 0.15, 500)

flux_array = np.zeros_like(x_front)
fem_array = np.zeros_like(x_front)
force_array = np.zeros_like(x_front)

for idx, x in enumerate(x_front):
    if x < 0.0:
        flux_array[idx] = 0.0
        fem_array[idx] = 0.0
        force_array[idx] = 0.0
    elif 0.0 <= x <= width:
        flux_array[idx] = B_field_z * length * x
        fem_array[idx] = -B_field_z * length * v_vel # Fem de movimiento
        # Fuerza F = I*l x B = (E/R)*l*B = - (B^2 * l^2 * v) / R
        force_array[idx] = -(B_field_z**2 * length**2 * v_vel) / R_loop
    else:
        flux_array[idx] = B_field_z * length * width
        fem_array[idx] = 0.0
        force_array[idx] = 0.0

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(9.5, 9.5))

# Flujo magnético
ax1.plot(x_front * 100, flux_array * 1e3, 'royalblue', lw=2.5, label=r'Flujo Magnético $\Phi_B$')
ax1.set_title("Flujo Magnético a través de la Espira Móvil vs. Posición del Frente", fontweight='bold')
ax1.set_xlabel("Posición del frente de la espira x (cm)")
ax1.set_ylabel("Flujo (mWb)")
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Fem inducida
ax2.plot(x_front * 100, fem_array, 'darkorange', lw=2.5, label=r'Fem Inducida $\mathcal{E}$')
ax2.set_title("Fuerza Electromotriz Inducida en la Espira vs. Posición", fontweight='bold')
ax2.set_xlabel("Posición del frente de la espira x (cm)")
ax2.set_ylabel("Fem (V)")
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True)

# Fuerza de frenado magnética (Lenz)
ax3.plot(x_front * 100, force_array * 1e3, 'crimson', lw=2.5, label='Fuerza Oponente (Lenz)')
ax3.set_title("Fuerza Magnética Retardadora vs. Posición (Freno Electromagnético)", fontweight='bold')
ax3.set_xlabel("Posición del frente de la espira x (cm)")
ax3.set_ylabel("Fuerza (mN)")
ax3.grid(True, ls='--', alpha=0.5)
ax3.legend(frameon=True)

plt.tight_layout()
plt.show()
```

## 9. Verificación Analítica con SymPy

Utilizaremos SymPy para plantear y resolver de forma exacta y simbólica la ecuación diferencial ordinaria de crecimiento de corriente en el circuito RL:
$$L \frac{di(t)}{dt} + R i(t) = \mathcal{E}$$
Sujeta a la condición inicial de corriente nula ($i(0) = 0$). Esto validará analíticamente la expresión de crecimiento exponencial.


```python
import sympy as sp

# Definir variables simbólicas reales
t_sym = sp.symbols('t', real=True, positive=True)
R_sym, L_sym, E_sym = sp.symbols('R L E', real=True, positive=True)
i_func = sp.Function('i')(t_sym)

# Plantear la EDO del transitorio RL
edo_rl = sp.Eq(L_sym * i_func.diff(t_sym) + R_sym * i_func, E_sym)

# Resolver con la condición inicial i(0) = 0
sol_edo = sp.dsolve(edo_rl, ics={i_func.subs(t_sym, 0): 0})
sol_simpl = sp.simplify(sol_edo.rhs)

# Comparación con el valor esperado teórico
val_esperado_rl = (E_sym / R_sym) * (1.0 - sp.exp(-R_sym * t_sym / L_sym))
diferencia_simb = sp.simplify(sol_simpl - val_esperado_rl)

print("======================================================================")
print("RESOLUCIÓN SIMBÓLICA DE LA EDO DEL CIRCUITO RL CON SYMPY")
print("======================================================================")
print(f"Ecuación Diferencial:         {edo_rl}")
print(f"Solución Resuelta Simbólica:  i(t) = {sol_simpl}")
print(f"Valor Esperado Teórico:       {val_esperado_rl}")
print(f"Diferencia simbólica:         {diferencia_simb} (Esperado: 0)")

assert diferencia_simb == 0, "Error: ¡La solución de la EDO no coincide con la teoría!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! La EDO se validó analíticamente de forma exacta.")
print("======================================================================")
```

## 1. Corriente de Desplazamiento de Maxwell

La Ley de Ampère original ($\oint \vec{B} \cdot d\vec{l} = \mu_0 I$) presentaba una inconsistencia lógica fundamental en circuitos con corrientes no estacionarias, como un condensador en proceso de carga o descarga. Si aplicamos la ley usando una superficie plana $S_1$ que atraviesa el cable de conducción, la corriente encerrada es la corriente real $I_c$ (corriente de conducción). Sin embargo, si deformamos la superficie a una forma de bolsa ($S_2$) que pase por la brecha vacía entre las placas del condensador, la corriente encerrada es cero (ya que no fluyen cargas por el vacío). Así, una misma trayectoria de integración de línea daría dos resultados incompatibles:
$$\oint \vec{B} \cdot d\vec{l} = \mu_0 I_c \quad \text{y} \quad \oint \vec{B} \cdot d\vec{l} = 0$$

Para corregir esta discontinuidad, Maxwell propuso que un campo eléctrico variable en el tiempo entre las placas del condensador actúa como una **fuente de campo magnético equivalente**, denominándolo **corriente de desplazamiento** $I_d$:
$$I_d = \epsilon_0 \frac{d\Phi_E}{dt}$$
Donde $\Phi_E = \int \vec{E} \cdot d\vec{A}$ es el flujo del campo eléctrico.

### 1.1 Ley de Ampère-Maxwell Modificada
Incorporando la corriente de desplazamiento, la ley modificada que unifica la generación de campos magnéticos por corrientes físicas y campos eléctricos variables es:
$$\oint \vec{B} \cdot d\vec{l} = \mu_0 \left(I_c + I_d\right) = \mu_0 \left(I_c + \epsilon_0 \frac{d\Phi_E}{dt}\right)$$


## 2. Ecuaciones de Maxwell en el Vacío

Las cuatro ecuaciones que unifican el electromagnetismo clásico describen el comportamiento e interrelación de los campos eléctrico $\vec{E}$ y magnético $\vec{B}$. En el vacío (espacio libre sin cargas ni corrientes libres, $Q_{enc} = 0$, $I_c = 0$), se enuncian como:

### 2.1 Ecuaciones en Forma Integral
1. **Ley de Gauss para Electricidad:** Relaciona el flujo eléctrico neto con la carga encerrada (cero en vacío):
   $$\oint \vec{E} \cdot d\vec{A} = 0$$
2. **Ley de Gauss para Magnetismo:** Postula la no existencia de monopolos magnéticos:
   $$\oint \vec{B} \cdot d\vec{A} = 0$$
3. **Ley de Faraday de la Inducción:** Campo magnético variable genera campo eléctrico:
   $$\oint \vec{E} \cdot d\vec{l} = -\frac{d\Phi_B}{dt}$$
4. **Ley de Ampère-Maxwell:** Campo eléctrico variable genera campo magnético (en vacío $I_c = 0$):
   $$\oint \vec{B} \cdot d\vec{l} = \mu_0 \epsilon_0 \frac{d\Phi_E}{dt}$$

### 2.2 Ecuaciones en Forma Diferencial
Utilizando los teoremas de Gauss (divergencia) y Stokes (rotacional), las ecuaciones diferenciales locales en el vacío son:
$$\nabla \cdot \vec{E} = 0 \qquad \nabla \cdot \vec{B} = 0$$
$$\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t} \qquad \nabla \times \vec{B} = \mu_0 \epsilon_0 \frac{\partial \vec{E}}{\partial t}$$

### 2.3 Ley de Fuerza de Lorentz
Estas ecuaciones describen los campos. La fuerza ejercida por estos campos sobre una partícula cargada de carga $q$ que se mueve a velocidad $\vec{v}$ viene descrita por la Ley de Fuerza de Lorentz:
$$\vec{F} = q\left(\vec{E} + \vec{v} \times \vec{B}\right)$$


## 3. Deducción Analítica de la Ecuación de Ondas

Consideremos una onda electromagnética plana que se propaga en el espacio libre a lo largo del eje positivo $x$. Supongamos que el campo eléctrico posee componentes únicamente en el eje $y$, $\vec{E} = E_y(x,t) \hat{j}$, y el campo magnético posee componentes únicamente en el eje $z$, $\vec{B} = B_z(x,t) \hat{k}$.

### 3.1 Aplicación de Ley de Faraday
Evaluando la Ley de Faraday en un rectángulo diferencial en el plano $xy$ con altura $a$ y anchura $\Delta x$:
$$\oint \vec{E} \cdot d\vec{l} = -E_y(x,t) a + E_y(x+\Delta x, t) a = a \left(E_y(x+\Delta x, t) - E_y(x,t)\right)$$
El flujo magnético a través del rectángulo es $\Phi_B = B_z(x,t) a \Delta x$. La variación temporal es:
$$\frac{d\Phi_B}{dt} = \frac{\partial B_z(x,t)}{\partial t} a \Delta x$$
Igualando por Faraday y dividiendo por $a \Delta x$ en el límite $\Delta x \to 0$:
$$\frac{\partial E_y}{\partial x} = -\frac{\partial B_z}{\partial t} \qquad (\text{Ec. A})$$

### 3.2 Aplicación de Ley de Ampère-Maxwell
Evaluando la ley en un rectángulo diferencial en el plano $xz$ de altura $a$ y anchura $\Delta x$:
$$\oint \vec{B} \cdot d\vec{l} = B_z(x,t) a - B_z(x+\Delta x, t) a = -a \left(B_z(x+\Delta x, t) - B_z(x,t)\right)$$
El flujo eléctrico es $\Phi_E = E_y(x,t) a \Delta x$. Su variación temporal es:
$$\frac{d\Phi_E}{dt} = \frac{\partial E_y(x,t)}{\partial t} a \Delta x$$
Igualando por Ampère-Maxwell y dividiendo por $a \Delta x$ en el límite $\Delta x \to 0$:
$$-\frac{\partial B_z}{\partial x} = \mu_0 \epsilon_0 \frac{\partial E_y}{\partial t} \qquad (\text{Ec. B})$$

### 3.3 Obtención de la Ecuación de Onda
Derivando la Ec. A respecto a $x$:
$$\frac{\partial^2 E_y}{\partial x^2} = -\frac{\partial^2 B_z}{\partial x \partial t}$$
Derivando la Ec. B respecto a $t$:
$$-\frac{\partial^2 B_z}{\partial t \partial x} = \mu_0 \epsilon_0 \frac{\partial^2 E_y}{\partial t^2}$$
Como las derivadas parciales cruzadas son idénticas ($\frac{\partial^2 B_z}{\partial x \partial t} = \frac{\partial^2 B_z}{\partial t \partial x}$), igualamos ambas expresiones y obtenemos la **Ecuación de Onda Unidimensional para el Campo Eléctrico**:
$$\frac{\partial^2 E_y}{\partial x^2} = \mu_0 \epsilon_0 \frac{\partial^2 E_y}{\partial t^2}$$\n
De manera análoga, derivando la Ec. A respecto a $t$ y la Ec. B respecto a $x$, obtenemos la ecuación de onda para el campo magnético:
$$\frac{\partial^2 B_z}{\partial x^2} = \mu_0 \epsilon_0 \frac{\partial^2 B_z}{\partial t^2}$$

### 3.4 La Velocidad de la Luz
La ecuación clásica de onda posee la estructura general $\frac{\partial^2 \Psi}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 \Psi}{\partial t^2}$. De esto se concluye que las perturbaciones electromagnéticas se propagan en el vacío con una rapidez de:
$$c = \frac{1}{\sqrt{\mu_0 \epsilon_0}}$$
Evaluando las constantes del vacío $\mu_0 = 4\pi \times 10^{-7}\text{ T}\cdot\text{m/A}$ y $\epsilon_0 = 8.85419 \times 10^{-12}\text{ C}^2/\text{N}\cdot\text{m}^2$, calculamos:
$$c = 2.99792458 \times 10^8\text{ m/s} \approx 3 \times 10^8\text{ m/s}$$
Esta concordancia teórica exacta demostró que **la luz es una onda electromagnética**.


## 4. Propiedades de las Ondas Electromagnéticas

A partir de las Ecuaciones de Maxwell se deducen las propiedades clave de las ondas electromagnéticas planas en el vacío:
1. **Son Transversales:** Los campos eléctrico $\vec{E}$ y magnético $\vec{B}$ son perpendiculares a la dirección de propagación de la onda.
2. **Ortogonalidad Mutua:** Los campos $\vec{E}$ y $\vec{B}$ son perpendiculares entre sí en todo punto ($\vec{E} \cdot \vec{B} = 0$).
3. **Dirección de Propagación:** Viene dada por el producto vectorial $\vec{E} \times \vec{B}$. El sentido del transporte de energía se describe mediante el vector de Poynting $\vec{S} = \frac{1}{\mu_0} \vec{E} \times \vec{B}$.
4. **Relación de Magnitudes:** En todo instante del tiempo y posición del espacio, las magnitudes de los campos cumplen de forma rígida la proporcionalidad:
   $$E = c B \qquad \text{o} \qquad \frac{E}{B} = c$$


## 5. Ondas Sinusoidales y Espectro Electromagnético

### 5.1 Ondas Electromagnéticas Sinusoidales
Las soluciones armónicas periódicas a la ecuación de onda se escriben como:
$$E_y(x,t) = E_{max} \cos(kx - \omega t)$$
$$B_z(x,t) = B_{max} \cos(kx - \omega t)$$
Donde:
* $k = \frac{2\pi}{\lambda}$ es el número de onda angular (rad/m), siendo $\lambda$ la longitud de onda.
* $\omega = 2\pi f$ es la frecuencia angular (rad/s), siendo $f$ la frecuencia de oscilación.
* La velocidad de fase de la onda cumple la relación de dispersión: $c = \frac{\omega}{k} = \lambda f$.

### 5.2 Análisis y Corrección de Errata de Rangos en la Tabla I (pág 18)

> [!IMPORTANT]
> **Errata Física en la Tabla I del Espectro:**
> La Tabla I en la página 18 del libro indica que las microondas tienen un rango de longitud de onda entre $1\text{ mm}$ y $1\text{ m}$, y escribe su frecuencia aproximada asociada como:
> $$f = 3 \times 10^{11} - 3 \times 10^{12}\text{ Hz}$$
> * **Corrección Física:** Esto es una inconsistencia dimensional grave con la relación de dispersión en el vacío $c = \lambda f$. Si realizamos la conversión rigurosa de las longitudes de onda límites:
>   * Para el límite superior de longitud de onda $\lambda = 1\text{ m}$, la frecuencia es:
>     $$f = \frac{c}{\lambda} = \frac{3 \times 10^8\text{ m/s}}{1\text{ m}} = 3 \times 10^8\text{ Hz} = 300\text{ MHz}$$
>   * Para el límite inferior de longitud de onda $\lambda = 1\text{ mm} = 10^{-3}\text{ m}$, la frecuencia es:
>     $$f = \frac{c}{\lambda} = \frac{3 \times 10^8\text{ m/s}}{10^{-3}\text{ m}} = 3 \times 10^{11}\text{ Hz} = 300\text{ GHz}$$
>   Por tanto, el rango físico correcto de frecuencia para las microondas es de $3 \times 10^8\text{ Hz} - 3 \times 10^{11}\text{ Hz}$ (300 MHz - 300 GHz). El libro de texto cometió una errata al indicar la frecuencia superior como $3 \times 10^{12}\text{ Hz}$ (que corresponde al infrarrojo lejano) y la inferior como $3 \times 10^{11}\text{ Hz}$ (el cual es el límite de alta frecuencia de la banda). Aclaramos este hecho y mostraremos las frecuencias calculadas de forma exacta en el desarrollo del cuaderno.


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: PROPAGACIÓN 3D DE ONDA ELECTROMAGNÉTICA PLANA ARMÓNICA
# =======================================================================
E_max = 5.0         # Amplitud de campo eléctrico (V/m)
c_light = 3e8       # Rapidez de la luz (m/s)
B_max = E_max / c_light # Amplitud de campo magnético (T)
freq = 1e9          # Frecuencia: 1 GHz
lam_val = c_light / freq # Longitud de onda: 0.3 m
k_num = 2.0 * np.pi / lam_val
omega_rad = 2.0 * np.pi * freq

# Espacio de propagación en X (de 0 a 2 longitudes de onda)
x_vals = np.linspace(0, 2 * lam_val, 350)
t_instant = 0.0     # Instante t = 0

# Magnitudes instantáneas de los campos cruzados
E_y = E_max * np.cos(k_num * x_vals - omega_rad * t_instant)
B_z = B_max * np.cos(k_num * x_vals - omega_rad * t_instant)

fig = plt.figure(figsize=(10.5, 8))
ax = fig.add_subplot(111, projection='3d')

# Línea de referencia del eje de propagación X
ax.plot(x_vals * 100, np.zeros_like(x_vals), np.zeros_like(x_vals), 'black', lw=1.5)

# Trazado de las curvas sinusoidales continuas de campo
ax.plot(x_vals * 100, E_y, np.zeros_like(x_vals), 'crimson', lw=2.2, label=r'Campo Eléctrico $\vec{E}(x)$')
# Escalamos B por la velocidad de la luz para poder graficarlo en la misma escala visual
ax.plot(x_vals * 100, np.zeros_like(x_vals), B_z * c_light, 'royalblue', lw=2.2, label=r'Campo Magnético $\vec{B}(x) \times c$')

# Dibujar vectores discretos de campo en puntos espaciados mediante quivers
quiver_step = 14
for idx in range(0, len(x_vals), quiver_step):
    # Vector de campo eléctrico (dirección Y)
    ax.quiver(x_vals[idx]*100, 0, 0, 0, E_y[idx], 0, color='crimson', arrow_length_ratio=0.15)
    # Vector de campo magnético (dirección Z)
    ax.quiver(x_vals[idx]*100, 0, 0, 0, 0, B_z[idx]*c_light, color='royalblue', arrow_length_ratio=0.15)

ax.set_title("Propagación de una Onda Electromagnética Plana Armónica en 3D", fontweight='bold')
ax.set_xlabel("Dirección de Propagación X (cm)")
ax.set_ylabel("Campo Eléctrico E_y (V/m)")
ax.set_zlabel("Campo Magnético B_z (T) Escalado")
ax.grid(True, ls='--', alpha=0.5)
ax.legend(frameon=True, loc='upper left')

# Ajuste de ángulo visual
ax.view_init(elev=22, azim=-52)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.patches as patches

# =======================================================================
# SIMULACIÓN 2: VISUALIZADOR INTERACTIVO DEL ESPECTRO ELECTROMAGNÉTICO
# =======================================================================

# Definimos las 7 regiones con sus rangos de longitud de onda límites en metros
spectrum_regions = [
    {"nombre": "Rayos Gamma", "lam_min": 1e-15, "lam_max": 1e-11, "color": "#4b0082"},
    {"nombre": "Rayos X", "lam_min": 1e-11, "lam_max": 10e-9, "color": "#8b008b"},
    {"nombre": "Ultravioleta", "lam_min": 10e-9, "lam_max": 400e-9, "color": "#9400d3"},
    {"nombre": "Visible", "lam_min": 400e-9, "lam_max": 700e-9, "color": "green"},
    {"nombre": "Infrarrojo", "lam_min": 700e-9, "lam_max": 1e-3, "color": "#ff4500"},
    {"nombre": "Microondas", "lam_min": 1e-3, "lam_max": 1.0, "color": "#ffd700"},
    {"nombre": "Ondas de Radio", "lam_min": 1.0, "lam_max": 1e5, "color": "#1e90ff"}
]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10.5, 9.0))

# Panel 1: Espectro general completo (eje logarítmico)
ax1.set_xscale('log')
ax1.set_ylim(0, 1.2)
ax1.set_xlim(1e-15, 1e5)

for reg in spectrum_regions:
    width = reg["lam_max"] - reg["lam_min"]
    rect = patches.Rectangle((reg["lam_min"], 0.2), width, 0.6, color=reg["color"], alpha=0.85)
    ax1.add_patch(rect)
    # Colocar la etiqueta en la media geométrica del rango
    x_center = np.sqrt(reg["lam_min"] * reg["lam_max"])
    ax1.text(x_center, 0.5, reg["nombre"], rotation=90, ha='center', va='center', 
             color='white', fontsize=10, fontweight='bold')

ax1.set_title("Espectro Electromagnético Completo (Longitud de Onda en el Vacío)", fontweight='bold')
ax1.set_xlabel(r"Longitud de Onda $\lambda$ (m)")
ax1.get_yaxis().set_visible(False)
ax1.grid(True, which="both", ls='--', alpha=0.5)

# Panel 2: Ventana detallada del espectro visible (lineal en nanómetros)
lam_vis = np.linspace(400.0, 700.0, 400)

# Función para mapear longitud de onda (nm) a color RGB aproximado para graficación
def wl_to_rgb(wl):
    w = float(wl)
    if 380 <= w < 440:
        R, G, B = -(w - 440) / (440 - 380), 0.0, 1.0
    elif 440 <= w < 490:
        R, G, B = 0.0, (w - 440) / (490 - 440), 1.0
    elif 490 <= w < 510:
        R, G, B = 0.0, 1.0, -(w - 510) / (510 - 490)
    elif 510 <= w < 580:
        R, G, B = (w - 510) / (580 - 510), 1.0, 0.0
    elif 580 <= w < 645:
        R, G, B = 1.0, -(w - 645) / (645 - 580), 0.0
    elif 645 <= w <= 780:
        R, G, B = 1.0, 0.0, 0.0
    else:
        R, G, B = 0.0, 0.0, 0.0
    
    # Atenuación en los límites de visibilidad del ojo humano
    factor = 1.0 if 420 <= w < 701 else (0.3 + 0.7*(w-380)/(420-380) if 380 <= w < 420 else (0.3 + 0.7*(780-w)/(780-701) if 701 <= w <= 780 else 0.0))
    return (R * factor, G * factor, B * factor)

colors_vis = [wl_to_rgb(w) for w in lam_vis]

# Trazar franjas verticales de colores espectrales continuos
for w_idx, wl in enumerate(lam_vis):
    ax2.axvline(wl, color=colors_vis[w_idx], lw=2.5)

ax2.set_xlim(400, 700)
ax2.set_ylim(0, 1.0)
ax2.get_yaxis().set_visible(False)
ax2.set_title("Detalle del Espectro Visible (Banda del Arcoíris)", fontweight='bold')
ax2.set_xlabel(r"Longitud de Onda $\lambda$ (nm)")
ax2.grid(True, ls='--', alpha=0.35)

plt.tight_layout()
plt.show()
```

## 7. Verificación Analítica con SymPy

Utilizaremos SymPy para verificar de forma simbólica la consistencia diferencial de la ecuación de onda. Definiremos la solución de prueba sinusoidal para el campo eléctrico:
$$E_y(x,t) = E_0 \cos(kx - \omega t)$$
Y evaluaremos el operador de la ecuación de onda tridimensional unidimensional en el vacío:
$$\frac{\partial^2 E_y}{\partial x^2} - \frac{1}{c^2} \frac{\partial^2 E_y}{\partial t^2}$$
Sustituyendo la relación de dispersión $\omega = c k$ para confirmar que el residuo diferencial es idénticamente cero.


```python
import sympy as sp

# Definir variables simbólicas reales
x_sym, t_sym, c_sym, E0_sym, k_sym, omega_sym = sp.symbols('x t c E_0 k omega', real=True)

# Solución sinusoidal de prueba
E_y_func = E0_sym * sp.cos(k_sym * x_sym - omega_sym * t_sym)

# Segundas derivadas parciales cruzadas
d2E_dx2 = sp.diff(E_y_func, x_sym, 2)
d2E_dt2 = sp.diff(E_y_func, t_sym, 2)

# Operador diferencial de la ecuación de onda
wave_operator = d2E_dx2 - (1.0 / c_sym**2) * d2E_dt2

# Sustituir la condición de dispersión omega = c * k
wave_simplified = sp.simplify(wave_operator.subs(omega_sym, c_sym * k_sym))

print("======================================================================")
print("VERIFICACIÓN DE LA ECUACIÓN DE ONDA ELECTROMAGNÉTICA CON SYMPY")
print("======================================================================")
print(f"Solución de Prueba E_y(x,t):     {E_y_func}")
print(f"Derivada espacial d^2E/dx^2:    {d2E_dx2}")
print(f"Derivada temporal d^2E/dt^2:    {d2E_dt2}")
print(f"Operador de Onda Planteado:      {wave_operator}")
print(f"Operador Simplificado (w = ck): {wave_simplified} (Esperado: 0)")

assert wave_simplified == 0, "Error: ¡La solución armónica falló la ecuación de onda!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Las funciones armónicas satisfacen la EDO exactamente.")
print("======================================================================")
```

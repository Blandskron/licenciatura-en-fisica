## 1. AnГЎlisis Vectorial Continuo: Operadores Diferenciales

En electromagnetismo trabajamos con dos entidades fundamentales que varГ­an en el espacio tridimensional:
* **Campos Escalares $\Phi(x, y, z)$:** Funciones que asignan un valor numГ©rico real a cada punto del espacio (ej. potencial elГ©ctrico $V$, temperatura $T$).
* **Campos Vectoriales $\vec{A}(x, y, z) = A_x\hat{i} + A_y\hat{j} + A_z\hat{k}$:** Funciones que asignan un vector (magnitud, direcciГіn y sentido) a cada punto del espacio (ej. campo elГ©ctrico $\vec{E}$, campo magnГ©tico $\vec{B}$, velocidad de un fluido $\vec{v}$).

Para estudiar cГіmo varГ­an estos campos localmente, introducimos el operador diferencial vectorial **Nabla** $\nabla$:
$$\nabla = \frac{\partial}{\partial x}\hat{i} + \frac{\partial}{\partial y}\hat{j} + \frac{\partial}{\partial z}\hat{k}$$

### 1.1 El Gradiente
Se aplica a un campo escalar $\Phi$, dando como resultado un campo vectorial:
$$\text{grad } \Phi = \nabla \Phi = \frac{\partial \Phi}{\partial x}\hat{i} + \frac{\partial \Phi}{\partial y}\hat{j} + \frac{\partial \Phi}{\partial z}\hat{k}$$
* **Significado FГ­sico:** El vector $\nabla \Phi$ apunta en la direcciГіn de mГЎximo crecimiento espacial del campo escalar $\Phi$, y su magnitud es la tasa de cambio en esa direcciГіn. En fГ­sica, las fuerzas conservativas se derivan del gradiente negativo de un potencial escalar, ej. $\vec{E} = -\nabla V$.

### 1.2 La Divergencia
Se aplica a un campo vectorial $\vec{A}$ mediante un producto escalar, dando como resultado un campo escalar:
$$\text{div } \vec{A} = \nabla \cdot \vec{A} = \frac{\partial A_x}{\partial x} + \frac{\partial A_y}{\partial y} + \frac{\partial A_z}{\partial z}$$
* **Significado FГ­sico:** La divergencia mide la densidad de flujo saliente neto de un campo vectorial por unidad de volumen alrededor de un punto. Si $\nabla \cdot \vec{A} > 0$, el punto actГєa como una **fuente** (ej. cargas elГ©ctricas positivas para el campo $\vec{E}$); si $\nabla \cdot \vec{A} < 0$, actГєa como un **sumidero** (cargas negativas); si $\nabla \cdot \vec{A} = 0$, el campo es **solenoidal** (sin fuentes netas, ej. el campo magnГ©tico $\vec{B}$).

### 1.3 El Rotacional
Se aplica a un campo vectorial $\vec{A}$ mediante un producto vectorial, dando como resultado otro campo vectorial:
$$\text{rot } \vec{A} = \nabla \times \vec{A} = \left(\frac{\partial A_z}{\partial y} - \frac{\partial A_y}{\partial z}\right)\hat{i} + \left(\frac{\partial A_x}{\partial z} - \frac{\partial A_z}{\partial x}\right)\hat{j} + \left(\frac{\partial A_y}{\partial x} - \frac{\partial A_x}{\partial y}\right)\hat{k}$$
* **Significado FГ­sico:** Mide la tendencia de un campo vectorial a circular o rotar localmente en torno a un punto. Si colocГЎramos una pequeГ±a hГ©lice en el campo, el eje de giro de la hГ©lice apuntarГ­a en la direcciГіn de $\nabla \times \vec{A}$, y su velocidad angular de rotaciГіn serГ­a proporcional a su magnitud. Si $\nabla \times \vec{A} = 0$, el campo es **irrotacional** o conservativo.

### 1.4 El Laplaciano
Es la divergencia del gradiente de un campo escalar, dando un campo escalar:
$$\nabla^2 \Phi = \nabla \cdot (\nabla \Phi) = \frac{\partial^2 \Phi}{\partial x^2} + \frac{\partial^2 \Phi}{\partial y^2} + \frac{\partial^2 \Phi}{\partial z^2}$$
Este operador es fundamental para formular la ecuaciГіn de Poisson y de Laplace en electrostГЎtica.


## 2. Teoremas Integrales Fundamentales

El cГЎlculo vectorial conecta la derivada local en un punto con sumas globales (integrales) a lo largo de volГєmenes o contornos. En electromagnetismo, dos teoremas actГєan como los puentes matemГЎticos para conectar el mundo microscГіpico diferencial con las leyes macroscГіpicas integrales medibles en laboratorio.

### 2.1 Teorema de Gauss (Divergencia)
Establece que el flujo neto de un campo vectorial $\vec{A}$ a travГ©s de una superficie cerrada $S$ es exactamente igual a la integral de volumen de su divergencia sobre el volumen $V$ encerrado por dicha superficie:
$$\oint_S \vec{A} \cdot d\vec{S} = \int_V \left(\nabla \cdot \vec{A}\right) dV$$
* **Importancia ElectromagnГ©tica:** Permite deducir la ley de Gauss en forma diferencial a partir de su versiГіn integral:
  $$\oint_S \vec{E} \cdot d\vec{S} = \frac{Q_{enc}}{\epsilon_0} = \int_V \frac{\rho}{\epsilon_0} dV \implies \int_V \left(\nabla \cdot \vec{E}\right) dV = \int_V \frac{\rho}{\epsilon_0} dV \implies \nabla \cdot \vec{E} = \frac{\rho}{\epsilon_0}$$

### 2.2 Teorema de Stokes
Establece que la circulaciГіn de un campo vectorial $\vec{A}$ a lo largo de una trayectoria cerrada curva $C$ es exactamente igual al flujo del rotacional de $\vec{A}$ a travГ©s de cualquier superficie abierta $S$ delimitada por dicha curva $C$:
$$\oint_C \vec{A} \cdot d\vec{l} = \int_S \left(\nabla \times \vec{A}\right) \cdot d\vec{S}$$
* **Importancia ElectromagnГ©tica:** Es el puente fundamental para deducir la forma diferencial de la Ley de Faraday y la Ley de AmpГЁre-Maxwell a partir de las integrales de lГ­nea de fem y fuerza magnetomotriz.


## 3. DiscretizaciГіn Espacial y MГ©todos de Rejilla (Grids)

En fГ­sica computacional, las computadoras no pueden evaluar funciones continuas sobre infinitos puntos del espacio. Por lo tanto, debemos dividir el dominio continuo en un conjunto finito de puntos espaciados de forma regular, creando una **malla o rejilla (grid)**.

Consideremos una malla cartesiana bidimensional estructurada, donde cada punto se identifica por sus Г­ndices $(i, j)$ tales que:
$$x_i = x_{min} + i \Delta x, \quad y_j = y_{min} + j \Delta y$$
donde $\Delta x$ y $\Delta y$ son los pasos de rejilla espaciales en los ejes X e Y respectivamente.

### 3.1 Derivadas por Diferencias Finitas
Podemos aproximar las derivadas parciales de un campo continuo evaluando las diferencias algebraicas discretas entre puntos adyacentes de la malla:
1. **Diferencia Centrada (Primer Orden - Gradiente):**
   Es la aproximaciГіn mГЎs precisa de primer orden, ya que posee un error de truncamiento cuadrГЎtico de orden $\mathcal{O}(\Delta x^2)$:
   $$\frac{\partial \Phi}{\partial x}\Big|_{(i,j)} \approx \frac{\Phi(x_{i+1}, y_j) - \Phi(x_{i-1}, y_j)}{2\Delta x}$$
2. **Diferencia Centrada de Segundo Orden (Laplaciano):**
   Utilizada para aproximar segundas derivadas parciales en ecuaciones diferenciales de segundo orden:
   $$\frac{\partial^2 \Phi}{\partial x^2}\Big|_{(i,j)} \approx \frac{\Phi(x_{i+1}, y_j) - 2\Phi(x_i, y_j) + \Phi(x_{i-1}, y_j)}{\Delta x^2}$$


````python
import numpy as np
import matplotlib.pyplot as plt

# ConfiguraciГіn estГ©tica de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIГ“N 1: VISUALIZACIГ“N 2D/3D DE CAMPOS ESCALARES Y VECTORIALES
# =======================================================================

# Definimos el Potencial de Carga electrostatico V(x, y) = 1 / sqrt(x^2 + y^2)
def potential_V(x, y):
    r = np.sqrt(x**2 + y**2)
    r[r < 1e-9] = 1e-9  # Evitar divisiГіn por cero en el origen
    return 1.0 / r

# Rango espacial para la malla
x_vals = np.linspace(-2.0, 2.0, 150)
y_vals = np.linspace(-2.0, 2.0, 150)
X, Y = np.meshgrid(x_vals, y_vals, indexing='ij')

# Evaluamos el campo escalar del potencial
V = potential_V(X, Y)

# Calculamos analГ­ticamente el campo elГ©ctrico asociado E = -grad V
# Ex = x / (x^2 + y^2)^(3/2), Ey = y / (x^2 + y^2)^(3/2)
r_dist = np.sqrt(X**2 + Y**2)
r_dist[r_dist < 1e-9] = 1e-9
E_x = X / (r_dist**3)
E_y = Y / (r_dist**3)

# GraficaciГіn en subplots
fig = plt.figure(figsize=(14, 6))

# Subplot 1: Superficie 3D del campo escalar (Potencial V)
ax1 = fig.add_subplot(121, projection='3d')
# Limitamos la visualizaciГіn en Z en el origen para evitar picos infinitos
V_capped = np.clip(V, 0.0, 4.0)
surf = ax1.plot_surface(X, Y, V_capped, cmap='plasma', edgecolor='none', alpha=0.95)
ax1.set_title("Superficie del Potencial Escalar $V(x,y)$", fontweight='bold')
ax1.set_xlabel("Eje X")
ax1.set_ylabel("Eje Y")
ax1.set_zlabel("Potencial V (V)")
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10, label='Voltaje (V)')
ax1.view_init(elev=35, azim=45)

# Subplot 2: Campo Vectorial E y curvas equipotenciales (2D Contour + Quiver)
ax2 = fig.add_subplot(122)
# Graficamos las curvas equipotenciales
contour = ax2.contour(X, Y, V_capped, levels=12, cmap='plasma', linewidths=1.8)
ax2.clabel(contour, inline=True, fontsize=9, fmt='%.2f')

# Graficamos las flechas del campo vectorial E. 
# Para evitar saturar el grГЎfico, mostramos 1 de cada 10 puntos en la malla.
skip = 10
# Normalizamos los vectores de E para visualizar solo la direcciГіn y sentido
E_mag = np.sqrt(E_x**2 + E_y**2)
E_mag[E_mag < 1e-9] = 1e-9
E_u = E_x / E_mag
E_v = E_y / E_mag

ax2.quiver(X[::skip, ::skip], Y[::skip, ::skip], E_u[::skip, ::skip], E_v[::skip, ::skip], 
          color='blue', alpha=0.6, width=0.003, scale=30, label=r'Campo ElГ©ctrico $\vec{E}$')

ax2.set_title("LГ­neas Equipotenciales y Vectores de Campo $\vec{E}$", fontweight='bold')
ax2.set_xlabel("Eje X")
ax2.set_ylabel("Eje Y")
ax2.set_xlim(-2.0, 2.0)
ax2.set_ylim(-2.0, 2.0)
ax2.legend(frameon=True, loc='upper right')
ax2.grid(True, ls='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

### 3.2 SimulaciГіn 2: ResoluciГіn NumГ©rica de Gradiente por Diferencias Finitas

Ahora programaremos de forma artesanal un algoritmo en diferencias finitas para calcular el gradiente numГ©rico del potencial $V(x,y)$ a lo largo de una rejilla regular de puntos y comparar la precisiГіn de este cГЎlculo numГ©rico frente al gradiente analГ­tico exacto.

Definiremos una malla en una regiГіn que excluya la singularidad en el origen, por ejemplo, el cuadrante positivo $x, y \in [0.5, 2.5]$, y calcularemos el error absoluto local medio:
$$\text{Error}_x(i,j) = \left| \frac{\partial V}{\partial x}\Big|_{analГ­tico} - \frac{\partial V}{\partial x}\Big|_{numГ©rico} \right|$$

````python
# =======================================================================
# SIMULACIГ“N 2: RESOLUTOR NUMГ‰RICO DE GRADIENTE EN REJILLAS
# =======================================================================

# Definimos la funciГіn de potencial y su gradiente analГ­tico exacto
def potential_V(x, y):
    return 1.0 / np.sqrt(x**2 + y**2)

def grad_V_analytical(x, y):
    r_3 = (x**2 + y**2)**1.5
    return -x / r_3, -y / r_3

# FunciГіn de diferencias finitas centradas
def grad_V_numerical(V_grid, dx, dy):
    dV_dx = np.zeros_like(V_grid)
    dV_dy = np.zeros_like(V_grid)
    
    # Diferencia centrada para puntos internos: (V[i+1, j] - V[i-1, j]) / (2*dx)
    dV_dx[1:-1, :] = (V_grid[2:, :] - V_grid[:-2, :]) / (2.0 * dx)
    dV_dy[:, 1:-1] = (V_grid[:, 2:] - V_grid[:, :-2]) / (2.0 * dy)
    
    # Diferencias de borde (hacia adelante/atrГЎs)
    dV_dx[0, :] = (V_grid[1, :] - V_grid[0, :]) / dx
    dV_dx[-1, :] = (V_grid[-1, :] - V_grid[-2, :]) / dx
    dV_dy[:, 0] = (V_grid[:, 1] - V_grid[:, 0]) / dy
    dV_dy[:, -1] = (V_grid[:, -1] - V_grid[:, -2]) / dy
    
    return dV_dx, dV_dy

# Malla numГ©rica regular en el cuadrante [0.5, 2.5]
n_points = 100
x_m = np.linspace(0.5, 2.5, n_points)
y_m = np.linspace(0.5, 2.5, n_points)
dx = x_m[1] - x_m[0]
dy = y_m[1] - y_m[0]
X_m, Y_m = np.meshgrid(x_m, y_m, indexing='ij')

# CГЎlculo
V_grid = potential_V(X_m, Y_m)
Ex_anal, Ey_anal = grad_V_analytical(X_m, Y_m)
Ex_num, Ey_num = grad_V_numerical(V_grid, dx, dy)

# Evaluamos errores absolutos en los puntos internos de la rejilla
error_Ex = np.abs(Ex_anal[1:-1, 1:-1] - Ex_num[1:-1, 1:-1])
error_Ey = np.abs(Ey_anal[1:-1, 1:-1] - Ey_num[1:-1, 1:-1])
mean_error_x = np.mean(error_Ex)
mean_error_y = np.mean(error_Ey)

# GraficaciГіn del mapa de error en el espacio
fig, ax = plt.subplots(figsize=(9, 6))
error_contour = ax.contourf(X_m[1:-1, 1:-1], Y_m[1:-1, 1:-1], error_Ex, cmap='viridis', levels=20)
fig.colorbar(error_contour, ax=ax, label='Error Absoluto de dV/dx')

ax.set_title("Mapa de Error Absoluto del Gradiente NumГ©rico (Diferencias Centradas)", fontweight='bold')
ax.set_xlabel("PosiciГіn X")
ax.set_ylabel("PosiciГіn Y")
ax.grid(True, ls='--', alpha=0.4)

plt.tight_layout()
plt.show()

print("======================================================================")
print("ANГЃLISIS DE ERROR Y PRECISIГ“N COMPUTACIONAL")
print("======================================================================")
print(f"Paso de Rejilla dx:                 {dx:.6f}")
print(f"Error Absoluto Medio en dV/dx:     {mean_error_x:.4e}")
print(f"Error Absoluto Medio en dV/dy:     {mean_error_y:.4e}")
print("======================================================================")
````

## 4. VerificaciГіn AnalГ­tica de Identidades Vectoriales con SymPy

Utilizaremos SymPy para realizar una verificaciГіn analГ­tica formal de dos identidades tridimensionales fundamentales del cГЎlculo vectorial que se utilizan de manera reiterada en el electromagnetismo clГЎsico:
1. **El rotacional del gradiente de cualquier campo escalar es idГ©nticamente cero:**
   $$\nabla \times \left(\nabla \Phi\right) = \vec{0}$$
2. **La divergencia del rotacional de cualquier campo vectorial es idГ©nticamente cero:**
   $$\nabla \cdot \left(\nabla \times \vec{A}\right) = 0$$

Ambas identidades se evaluarГЎn simbГіlicamente mediante el mГіdulo `sympy.vector` definiendo funciones diferenciables genГ©ricas.

````python
import sympy as sp
from sympy.vector import CoordSys3D, gradient, curl, divergence

# Definimos el sistema de coordenadas tridimensional cartesiano de SymPy
N = CoordSys3D('N')

# =======================================================================
# Identidad 1: curl(grad(Phi)) = 0
# =======================================================================
# Definimos una funciГіn escalar genГ©rica Phi(x, y, z)
Phi = sp.Function('Phi')(N.x, N.y, N.z)

# Calculamos el gradiente
grad_Phi = gradient(Phi)
# Calculamos el rotacional del gradiente
curl_grad_Phi = curl(grad_Phi)

# =======================================================================
# Identidad 2: div(curl(A)) = 0
# =======================================================================
# Definimos componentes de un campo vectorial genГ©rico A(x, y, z)
Ax = sp.Function('A_x')(N.x, N.y, N.z)
Ay = sp.Function('A_y')(N.x, N.y, N.z)
Az = sp.Function('A_z')(N.x, N.y, N.z)
A_vec = Ax * N.i + Ay * N.j + Az * N.k

# Calculamos el rotacional de A
curl_A = curl(A_vec)
# Calculamos la divergencia del rotacional
div_curl_A = divergence(curl_A)

# Resultados de la verificaciГіn simbГіlica
print("======================================================================")
print("VERIFICACIГ“N DE IDENTIDADES VECTORIALES CON SYMPY")
print("======================================================================")
print(f"Identidad 1: rot(grad(Phi)) = {curl_grad_Phi} (Esperado: 0*N.i + 0*N.j + 0*N.k)")
print(f"Identidad 2: div(rot(A))    = {div_curl_A} (Esperado: 0)")
print("======================================================================")

# Validamos mediante aserciones
assert curl_grad_Phi == 0 * N.i + 0 * N.j + 0 * N.k, "Error: ВЎrot(grad(Phi)) no es cero!"
assert div_curl_A == 0, "Error: ВЎdiv(rot(A)) no es cero!"
print("ВЎVERIFICACIГ“N COMPLETA! Las identidades diferenciales vectoriales son correctas.")
print("======================================================================")
````

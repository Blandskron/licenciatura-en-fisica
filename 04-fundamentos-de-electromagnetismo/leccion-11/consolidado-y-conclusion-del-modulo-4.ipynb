## 2. Propiedades Magnéticas de la Materia

Cuando un material se introduce en un campo magnético externo $\vec{B}_0$, los electrones en sus átomos (debido a su movimiento orbital y a su momento magnético de espín) interactúan con el campo. Esta respuesta microscópica se describe macroscópicamente mediante el vector **Magnetización** $ec{M}$, que representa el momento magnético dipolar neto por unidad de volumen.

El campo magnético total $ec{B}$ en el interior del material es la suma del campo que existiría en el vacío debido a corrientes libres (descrito por el campo auxiliar $ec{H}$) y la contribución de los momentos dipolares del medio:
$$\vec{B} = \mu_0 \left(\vec{H} + \vec{M}\right)$$

En medios materiales lineales, la magnetización es directamente proporcional a la intensidad del campo magnético $ec{H}$:
$$\vec{M} = \chi_m \vec{H}$$
Donde $\chi_m$ es una cantidad adimensional denominada **susceptibilidad magnética**.

Sustituyendo esta relación:
$$\vec{B} = \mu_0 \left(1 + \chi_m\right) \vec{H} = \mu_0 \mu_r \vec{H} = \mu \vec{H}$$
Donde $\mu_r = 1 + \chi_m$ es la permeabilidad relativa y $\mu = \mu_0 \mu_r$ es la permeabilidad magnética del medio.

### 2.1 Clasificación del Comportamiento Magnético

1. **Diamagnetismo:**
   * **Origen:** Es un efecto puramente cuántico inducido en todos los átomos. De acuerdo con la ley de Lenz a escala microscópica, el campo externo altera el movimiento orbital de los electrones, induciendo un momento magnético neto opuesto al campo aplicado.
   * **Características:** La susceptibilidad magnética es negativa, muy pequeña ($\chi_m \sim -10^{-5}$) e independiente de la temperatura. Ejemplos: cobre, bismuto, agua.
2. **Paramagnetismo:**
   * **Origen:** Se presenta en materiales con átomos que poseen momentos magnéticos permanentes (electrones desapareados), pero orientados al azar debido a la agitación térmica. Al aplicar un campo externo, los dipolos tienden a alinearse con el campo, compitiendo contra la aleatorización térmica.
   * **Características:** La susceptibilidad es positiva, pequeña ($\chi_m \sim 10^{-5} - 10^{-3}$) y depende fuertemente de la temperatura. Su comportamiento se modela mediante la función de Langevin.
3. **Ferromagnetismo:**
   * **Origen:** Efecto colectivo de origen mecánico-cuántico (interacción de intercambio) que alinea fuertemente los espines de átomos vecinos en regiones localizadas llamadas **dominios magnéticos**, incluso en ausencia de un campo externo.
   * **Características:** Posee una susceptibilidad extremadamente grande y no lineal ($\chi_m \sim 10^3 - 10^5$), exhibiendo el fenómeno de histéresis (memoria magnética) y perdiendo sus propiedades ferromagnéticas por encima de la **temperatura de Curie** $T_C$, convirtiéndose en paramagnético. Ejemplos: hierro, cobalto, níquel.


### 2.2 Paramagnetismo y la Función de Langevin

Desde una perspectiva de mecánica estadística clásica, consideremos un sistema de dipolos magnéticos con momento dipolar $\vec{\mu}$ en presencia de un campo magnético $\vec{B}$. La energía potencial de cada dipolo es $U = -\vec{\mu} \cdot \vec{B} = -\mu B \cos(\theta)$.

La distribución de Boltzmann establece que la probabilidad de encontrar un dipolo en un ángulo $\theta$ es proporcional a $e^{-U/(k_B T)} = e^{\mu B \cos(\theta)/(k_B T)}$. Al promediar la componente de la magnetización paralela al campo sobre todos los ángulos en tres dimensiones, se obtiene la **Magnetización de Langevin**:
$$M = M_{sat} L(x)$$
Donde $M_{sat} = N \mu$ es la magnetización de saturación (cuando todos los dipolos están perfectamente alineados), y la **función de Langevin** $L(x)$ se define como:
$$L(x) = \coth(x) - \frac{1}{x}$$
con el parámetro adimensional:
$$x = \frac{\mu B}{k_B T}$$
El cual relaciona la energía de interacción magnética con la energía térmica media.


````python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: MODELADO DE LA FUNCIÓN DE LANGEVIN Y LEY DE CURIE
# =======================================================================

def langevin_func(x):
    """
    Evalúa la función de Langevin L(x) = coth(x) - 1/x.
    Para valores de x cercanos a 0, se utiliza su desarrollo en serie de Taylor
    L(x) ≈ x/3 - x^3/45 para evitar la indeterminación numérica.
    """
    x = np.asarray(x, dtype=float)
    # Creamos una máscara para valores muy pequeños de x
    small_x = np.abs(x) < 1e-4
    
    result = np.zeros_like(x)
    # Para x pequeño: aproximación de Taylor
    result[small_x] = x[small_x] / 3.0 - (x[small_x]**3) / 45.0
    # Para x general: fórmula exacta
    non_small = ~small_x
    result[non_small] = 1.0 / np.tanh(x[non_small]) - 1.0 / x[non_small]
    return result

# Rango del parámetro x = mu*B / (kB*T)
x_vals = np.linspace(-10, 10, 500)
L_vals = langevin_func(x_vals)

# Aproximación lineal de Curie para campos débiles o altas temperaturas (x << 1)
x_linear = np.linspace(-2, 2, 100)
L_linear = x_linear / 3.0

# Graficación
fig, ax = plt.subplots(figsize=(9, 5.5))
ax.plot(x_vals, L_vals, 'crimson', lw=2.5, label='Función de Langevin Exacta $L(x)$')
ax.plot(x_linear, L_linear, 'royalblue', ls='--', lw=2.0, label='Ley de Curie (Límite Lineal $x/3$)')
ax.axhline(1.0, color='gray', ls=':', label='Saturación $L(x) \to \pm 1$')
ax.axhline(-1.0, color='gray', ls=':')
ax.axvline(0, color='black', lw=0.8)
ax.axhline(0, color='black', lw=0.8)

ax.set_title("Magnetización de Langevin en Paramagnetos", fontweight='bold')
ax.set_xlabel(r"Parámetro Adimensional $x = \mu B / (k_B T)$")
ax.set_ylabel(r"Magnetización Relativa $M / M_{sat}$")
ax.set_xlim(-10, 10)
ax.set_ylim(-1.25, 1.25)
ax.legend(frameon=True, loc='upper left')
ax.grid(True, ls='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

### 2.3 Ferromagnetismo e Histéresis Magnética

En un material ferromagnético, la relación $\vec{B}(\vec{H})$ o $\vec{M}(\vec{H})$ no es lineal ni biunívoca; depende de la historia magnética del material. Si partimos de un material completamente desmagnetizado ($\vec{H}=0$, $\vec{M}=0$) y aumentamos el campo externo $H$, la magnetización sigue la **curva de magnetización inicial** hasta alcanzar la saturación $M_{sat}$.

Si luego disminuimos el campo $H$ hasta cero, la magnetización no vuelve a cero, sino que retiene un valor denominado **magnetización de remanencia** $M_r$ (imán permanente). Para reducir la magnetización a cero, es necesario aplicar un campo en sentido opuesto de magnitud $H_c$, llamado **campo coercitivo**. Este comportamiento cíclico se conoce como **lazo de histéresis**. El área encerrada por el lazo representa la pérdida de energía por unidad de volumen en forma de calor durante un ciclo completo de magnetización.

Para modelar computacionalmente el comportamiento del ciclo de histéresis ferromagnética, implementaremos un modelo fenomenológico basado en funciones de tipo tangente hiperbólica ($\tanh$) desplazadas. Esto nos permite simular la transición entre la magnetización ascendente y descendente bajo un campo magnético oscilante $H(t) = H_{max} \cos(\omega t)$.

El modelo para las dos ramas se escribe como:
* **Rama Ascendente (campo $H$ aumentando):**
  $$M_{asc}(H) = M_{sat} \tanh\left(\frac{H - H_c}{a}\right) + \chi_0 H$$
* **Rama Descendente (campo $H$ disminuyendo):**
  $$M_{desc}(H) = M_{sat} \tanh\left(\frac{H + H_c}{a}\right) + \chi_0 H$$

Donde $M_{sat}$ es la magnetización de saturación, $H_c$ es el campo coercitivo, $a$ es un parámetro de forma y $\chi_0$ representa una susceptibilidad residual de fondo.


````python
# =======================================================================
# SIMULACIÓN 2: LAZO DE HISTÉRESIS FERROMAGNÉTICA NO LINEAL
# =======================================================================

# Parámetros del material (hierro dulce típico)
M_sat = 1.5         # Magnetización de saturación (Tesla)
H_c = 120.0         # Campo coercitivo (A/m)
shape_a = 150.0     # Parámetro de forma del lazo
chi_0 = 0.0005      # Susceptibilidad de fondo

# Generamos un campo magnético oscilante completo (ciclo de histéresis)
theta = np.linspace(0, 2 * np.pi, 600)
H_vals = 400.0 * np.cos(theta)  # Campo H variando entre -400 y 400 A/m

# Separamos los puntos en las ramas ascendente y descendente
M_vals = np.zeros_like(H_vals)

for i, th in enumerate(theta):
    H = H_vals[i]
    # Usamos el signo de sin(th) para determinar el sentido de cambio de H
    if np.sin(th) > 0:
        # Rama Descendente (dH/dt <= 0)
        M_vals[i] = M_sat * np.tanh((H + H_c) / shape_a) + chi_0 * H
    else: 
        # Rama Ascendente (dH/dt > 0)
        M_vals[i] = M_sat * np.tanh((H - H_c) / shape_a) + chi_0 * H

# Remanencia teórica M_r = M_sat * tanh(H_c / a)
M_r = M_sat * np.tanh(H_c / shape_a)

# Graficación del lazo de histéresis
fig, ax = plt.subplots(figsize=(9, 5.5))
ax.plot(H_vals, M_vals, 'forestgreen', lw=2.5, label='Lazo de Histéresis Ferromagnética')

# Marcadores de puntos clave
ax.plot(0, M_r, 'ro', ms=8, label=r'Magnetización de Remanencia $M_r$')
ax.plot(0, -M_r, 'ro', ms=8)
ax.plot(H_c, 0, 'ko', ms=8, label=r'Campo Coercitivo $H_c$')
ax.plot(-H_c, 0, 'ko')

# Estilo y anotaciones
ax.axhline(0, color='black', lw=0.8)
ax.axvline(0, color='black', lw=0.8)
ax.set_title("Ciclo de Histéresis del Hierro Dulce", fontweight='bold')
ax.set_xlabel(r"Intensidad de Campo Magnético $H$ (A/m)")
ax.set_ylabel(r"Magnetización $M$ (T)")
ax.set_xlim(-450, 450)
ax.set_ylim(-1.8, 1.8)
ax.grid(True, ls='--', alpha=0.6)
ax.legend(frameon=True, loc='upper left')

# Flechas indicadoras del sentido de recorrido
ax.annotate('', xy=(-200, 1.0), xytext=(-100, 1.2),
            arrowprops=dict(arrowstyle="->", color='darkgreen', lw=2.0))
ax.annotate('', xy=(200, -1.0), xytext=(100, -1.2),
            arrowprops=dict(arrowstyle="->", color='darkgreen', lw=2.0))

plt.tight_layout()
plt.show()
````

## 3. Estudio de Caso Integrador: Freno Magnético por Corrientes de Foucault (Eddy Currents)

Para unificar las leyes de la electrodinámica clásica estudiadas en este módulo, analizaremos y modelaremos el funcionamiento físico de un **freno magnético por corrientes de Foucault**. Este sistema acopla de manera elegante la inducción de Faraday, las corrientes de conducción en un medio material conductor y la fuerza magnética de Lorentz, resultando en un torque amortiguador puramente electromagnético.

### 3.1 Planteamiento Físico del Sistema

Consideremos un disco conductor de radio $R$, espesor $d$, y conductividad eléctrica $\sigma$, que gira libremente alrededor de su eje central con una velocidad angular inicial $\omega_0$. El disco posee un momento de inercia rotacional $I_{rot}$.

Un electroimán genera un campo magnético uniforme $ec{B}_0 = B_0 \hat{k}$ perpendicular a la superficie del disco. Este campo está localizado en una pequeña área rectangular $A = a 	imes b$ situada a una distancia radial media $r$ del eje del disco.

```
                  Disco Conductor Giratorio (Conductividad σ)
                             ┌───────────────┐
                             │       _..---..│_
                             │   .-''    │   │ ``-.
                             │  /        │   │     \
                             │ /         │   │      \
                             ││   (Axis) ┼───┼─► r  ││  ▲ Giro ω(t)
                             ││          │   │      ││  │
                             │ \         │  [Area A]│/
                             │  \        │  [ B0 ]  /
                             │   `-.     │     _.-'
                             │      ```--┬--'''
                             └───────────┘
```

### 3.2 Deducción del Torque de Frenado

Cuando el disco gira con velocidad angular $\omega(t)$, el metal en la región del campo magnético se desplaza con una velocidad lineal local:
$$v = \omega r$$

De acuerdo con la **Ley de Faraday** (Lección 8), el movimiento de este conductor a través del campo magnético induce una fuerza electromotriz (fem) de movimiento $\mathcal{E}$ a lo largo de la sección radial de la zona del polo:
$$\mathcal{E} \approx v B_0 b = \omega r B_0 b$$
Donde $b$ es la longitud radial del polo magnético.

Esta fem induce **corrientes de Foucault (corrientes parásitas)** en el cuerpo del disco conductor. De acuerdo con la Ley de Ohm en medios continuos (Lección 5), la densidad de corriente es $\vec{J} = \sigma \vec{E}$. Podemos modelar la resistencia eléctrica efectiva de la trayectoria cerrada de la corriente en el disco como proporcional a la resistividad del material y a factores geométricos:
$$R_{eff} \approx \frac{1}{\sigma d}$$
Donde la corriente inducida neta que atraviesa el campo magnético es:
$$I_{ind} \approx \frac{\mathcal{E}}{R_{eff}} \approx \sigma d \omega r B_0 b$$

De acuerdo con la **Ley de Fuerza de Lorentz** (Lección 6), el campo magnético $\vec{B}_0$ ejerce una fuerza magnética retardadora sobre esta corriente inducida. La magnitud de esta fuerza retardadora es:
$$F_B = I_{ind} b B_0 \approx \sigma d b^2 B_0^2 r \omega$$

Esta fuerza actúa en dirección opuesta a la velocidad lineal (debido a la **Ley de Lenz**), generando un torque de frenado o torque magnético $\tau_B$ opuesto al giro del disco:
$$\tau_B = -r F_B = - \left(\sigma d A r^2 B_0^2\right) \omega = - \alpha \omega$$
Donde hemos agrupado todos los parámetros físicos y geométricos en la constante de amortiguamiento magnético $\alpha$:
$$\alpha = \sigma d A r^2 B_0^2$$

### 3.3 Dinámica del Frenado y Conservación de la Energía

La ecuación de movimiento rotacional para el disco es:
$$I_{rot} \frac{d\omega}{dt} = \tau_B = - \alpha \omega$$

Esta es una ecuación diferencial ordinaria (EDO) lineal de primer orden (análoga al decaimiento de corriente en un circuito RL, Lección 8). Su solución analítica con condición inicial $\omega(0) = \omega_0$ es:
$$\omega(t) = \omega_0 e^{-t / \tau_f}$$
Donde la **constante de tiempo de frenado** $\tau_f$ es:
$$\tau_f = \frac{I_{rot}}{\alpha} = \frac{I_{rot}}{\sigma d A r^2 B_0^2}$$

La energía cinética rotacional instantánea del disco es:
$$E_c(t) = \frac{1}{2} I_{rot} \omega(t)^2 = \frac{1}{2} I_{rot} \omega_0^2 e^{-2t / \tau_f} = E_{c0} e^{-2t / \tau_f}$$

La potencia disipada por efecto Joule (calor) debido a las corrientes inducidas circulando por la resistencia del disco conductor es:
$$P_{dis}(t) = I_{ind}^2 R_{eff} = \tau_B \omega = \alpha \omega(t)^2 = \alpha \omega_0^2 e^{-2t / \tau_f}$$

Podemos comprobar la conservación de la energía verificando que la potencia disipada es exactamente igual a la tasa de pérdida de energía cinética del disco:
$$P_{dis}(t) = -\frac{dE_c(t)}{dt} = -\frac{d}{dt}\left(\frac{1}{2} I_{rot} \omega_0^2 e^{-2t / \tau_f}\right) = \frac{I_{rot}}{\tau_f} \omega_0^2 e^{-2t / \tau_f} = \alpha \omega_0^2 e^{-2t / \tau_f}$$
Esto confirma matemáticamente la conservación de la energía en el sistema.


````python
# =======================================================================
# SIMULACIÓN 3: DINÁMICA DE FRENADO Y DISIPACIÓN DE POTENCIA POR JOULE
# =======================================================================

# Parámetros Físicos y Geométricos (Cobre y Electroimán)
R_disk = 0.15          # Radio del disco de cobre (m)
thickness = 0.003      # Espesor del disco (m)
sigma_copper = 5.8e7   # Conductividad eléctrica del cobre (S/m)
density_copper = 8960.0 # Densidad del cobre (kg/m^3)

# Cálculo de la masa y momento de inercia del disco (I = 0.5 * M * R^2)
vol_disk = np.pi * (R_disk**2) * thickness
mass_disk = vol_disk * density_copper
I_rot = 0.5 * mass_disk * (R_disk**2)

# Parámetros del electroimán
B_0 = 0.15             # Intensidad del campo magnético localizado (T)
A_pole = 4.0e-4        # Área de la cara del polo magnético (2 cm x 2 cm = 4e-4 m^2)
r_mean = 0.10          # Distancia radial media del polo al eje (m)

# Parámetros del Amortiguamiento y Constante de Tiempo
# alpha = sigma * d * A * r^2 * B_0^2
alpha = sigma_copper * thickness * A_pole * (r_mean**2) * (B_0**2)
tau_f = I_rot / alpha  # Constante de tiempo de frenado (s)

# Velocidad angular inicial (en rad/s y RPM)
omega_0 = 120.0        # rad/s (~1145 RPM)
omega_0_rpm = omega_0 * 30 / np.pi

# Simulación temporal
t_max = 5 * tau_f      # Simular hasta 5 constantes de tiempo
t_array = np.linspace(0, t_max, 500)

# Solución analítica para la velocidad angular y energía cinética
omega_t = omega_0 * np.exp(-t_array / tau_f)
E_kinetic = 0.5 * I_rot * (omega_t**2)

# Potencia disipada instantánea por corrientes de Foucault (P = alpha * omega^2)
P_dissipated = alpha * (omega_t**2)
# Tasa de pérdida de energía cinética calculada de forma numérica (dEc/dt)
dEc_dt_num = -np.diff(E_kinetic) / np.diff(t_array)
t_mid = t_array[:-1] + np.diff(t_array)/2.0

# Graficación en Subplots
fig, ax1 = plt.subplots(figsize=(10.5, 6))

# Panel izquierdo/principal: Velocidad angular y energía cinética
color = 'crimson'
ax1.set_xlabel('Tiempo $t$ (s)', fontweight='bold')
ax1.set_ylabel(r'Velocidad Angular $\omega(t)$ (rad/s)', color=color, fontweight='bold')
line1 = ax1.plot(t_array, omega_t, color=color, lw=2.5, label=r'Velocidad Angular $\omega(t)$')
ax1.tick_params(axis='y', labelcolor=color)
ax1.axvline(tau_f, color='black', ls='--', alpha=0.7, label=r'Constante de tiempo $\tau_f$')

ax1_twin = ax1.twinx()
color = 'royalblue'
ax1_twin.set_ylabel('Energía Cinética $E_c(t)$ (J)', color=color, fontweight='bold')
line2 = ax1_twin.plot(t_array, E_kinetic, color=color, lw=2.5, ls='-.', label='Energía Cinética $E_c(t)$')
ax1_twin.tick_params(axis='y', labelcolor=color)

# Combinar leyendas de ambos ejes
lines = line1 + line2 + [plt.Line2D([0], [0], color='black', linestyle='--')]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper right', frameon=True)
ax1.set_title("Dinámica de Frenado del Disco de Cobre y Balance de Energía", fontweight='bold')
ax1.grid(True, ls='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Gráfico secundario: Verificación del balance térmico
fig2, ax2 = plt.subplots(figsize=(10.5, 5))
ax2.plot(t_array, P_dissipated, 'darkorange', lw=3.0, label=r'Potencia Disipada Joule $P_{dis} = \alpha \omega^2$')
ax2.plot(t_mid, dEc_dt_num, 'black', ls='--', lw=1.8, label=r'Pérdida de Energía Cinética $-dE_c/dt$')
ax2.set_xlabel('Tiempo $t$ (s)', fontweight='bold')
ax2.set_ylabel('Potencia $P$ (W)', fontweight='bold')
ax2.set_title("Verificación de la Conservación de Energía (Térmica vs. Mecánica)", fontweight='bold')
ax2.legend(frameon=True, loc='upper right')
ax2.grid(True, ls='--', alpha=0.5)

param_text = (
    f"Propiedades del Sistema:\n"
    f" Masa disco: {mass_disk:.2f} kg\n"
    f" Inercia I: {I_rot:.5f} kg·m²\n"
    f" Amortiguamiento α: {alpha:.5f} N·m·s\n"
    f" Constante tiempo τf: {tau_f:.3f} s\n"
    f" Velocidad inicial: {omega_0_rpm:.0f} RPM"
)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.4)
ax2.text(0.40, 0.45, param_text, transform=ax2.transAxes, fontsize=10, bbox=props, verticalalignment='top')
plt.tight_layout()
plt.show()
````

## 4. Verificación Analítica con SymPy

Utilizaremos SymPy para realizar una verificación analítica formal del comportamiento de la magnetización paramagnética en el límite de campos débiles o altas temperaturas ($x \ll 1$).

Definimos la función de Langevin de manera simbólica:
$$L(x) = \coth(x) - \frac{1}{x} = \frac{\cosh(x)}{\sinh(x)} - \frac{1}{x}$$

Calcularemos su desarrollo en series de Taylor en torno a $x=0$ hasta el término de quinto orden para validar que el término de primer orden dominante es $\frac{x}{3}$, lo que conduce de forma matemática rigurosa a la **Ley de Curie**:
$$M \approx M_{sat} \frac{x}{3} = N \mu \frac{\mu B}{3 k_B T} = \left(\frac{N \mu^2}{3 k_B}\right) \frac{B}{T} = C \frac{B}{T}$$
Donde $C = \frac{N \mu^2}{3 k_B}$ es la constante de Curie.


````python
import sympy as sp

# Definimos la variable simbólica real
x_sym = sp.symbols('x', real=True)

# Planteamos la función de Langevin simbólica
L_sym = sp.cosh(x_sym) / sp.sinh(x_sym) - 1 / x_sym

# Calculamos el desarrollo en serie de Taylor en x=0 hasta orden 5
L_series = sp.series(L_sym, x_sym, 0, 5)

# Extraemos el término principal
lead_term = L_series.as_leading_term(x_sym)

print("======================================================================")
print("VERIFICACIÓN ANALÍTICA DE LA LEY DE CURIE CON SYMPY")
print("======================================================================")
print(f"Función de Langevin Simbólica L(x):   {L_sym}")
print(f"Desarrollo en serie de Taylor (x->0):  {L_series}")
print(f"Término de primer orden dominante:    {lead_term}  (Esperado: x/3)")
print("======================================================================")

# Verificación automática mediante aserción
assert lead_term == x_sym / 3, "Error: ¡El término dominante de Langevin no coincide con x/3!"
print("¡VERIFICACIÓN COMPLETA! Se ha demostrado analíticamente la Ley de Curie.")
print("======================================================================")
````

## 1. Magnetismo y Campo Magnético

El magnetismo se observó originalmente en imanes permanentes naturales de magnetita ($Fe_3O_4$). Al igual que la carga eléctrica, la interacción magnética se describe mediante un campo, en este caso el **campo magnético** (indución magnética) $\vec{B}$, que es una magnitud vectorial.

### 1.1 Polos Magnéticos y Ausencia de Monopolos
Todo imán posee dos polos magnéticos denominados **Norte (N)** y **Sur (S)**:
* **Atracción y Repulsión:** Polos del mismo tipo se repelen; polos opuestos se atraen.
* **Indivisibilidad (Inexistencia de Monopolos):** A diferencia de las cargas eléctricas (donde existen electrones y protones aislados), los polos magnéticos siempre vienen en parejas. Si cortamos un imán de barra por la mitad, no obtenemos monopolos aislados, sino dos imanes bipolares más pequeños. Hasta el momento no se ha detectado experimentalmente la existencia de un monopolo magnético.

### 1.2 Líneas de Campo Magnético
Las líneas de campo magnético salen de los polos Norte y entran en los polos Sur de forma externa, cerrando sus trayectorias por el interior del imán (son curvas cerradas continuas, lo que se traduce en que la divergencia del campo magnético es cero: $\nabla \cdot \vec{B} = 0$).

### 1.3 Magnetismo Terrestre
La Tierra posee su propio campo magnético dipolar. Debido a que el polo norte de una brújula apunta hacia el norte geográfico de la Tierra, el polo norte geográfico es magnéticamente un **polo sur magnético**, y viceversa.


## 2. Fuerza de Lorentz sobre una Carga Puntual

El campo magnético no ejerce fuerza sobre cargas estáticas. La fuerza magnética $\vec{F}_B$ actúa únicamente sobre una carga de prueba $q$ en movimiento con velocidad $\vec{v}$ dentro del campo:
$$\vec{F}_B = q \vec{v} \times \vec{B}$$
La magnitud de la fuerza viene dada por:
$$F_B = |q| v B \sin\theta$$
Donde $\theta$ es el menor ángulo entre la velocidad $\vec{v}$ y el campo $\vec{B}$.

### 2.1 Propiedades de la Fuerza Magnética
1. **Dirección Perpendicular:** Al ser un producto vectorial, $\vec{F}_B$ es ortogonal tanto a la velocidad $\vec{v}$ como al campo $\vec{B}$. La dirección se determina mediante la **regla de la mano derecha**.
2. **Dependencia del Signo:** Si la carga es negativa, la dirección de la fuerza es opuesta a la obtenida con la regla de la mano derecha.
3. **Fuerza Nula en Movimiento Paralelo:** Si $\theta = 0^\circ$ o $180^\circ$ (movimiento paralelo o antiparalelo), la fuerza es estrictamente cero.
4. **Unidades:** En el SI la magnitud del campo magnético se mide en **Teslas (T)**:
   $$1\text{ T} = 1\text{ }\frac{\text{N}}{\text{C} \cdot \text{m/s}} = 1\text{ }\frac{\text{N}}{\text{A} \cdot \text{m}}$$
   Fuera del SI se usa el **Gauss (G)** ($1\text{ T} = 10^4\text{ G}$).

### 2.2 Análisis y Corrección de Erratas del Libro Guía

> [!IMPORTANT]
> **Errata 1: Exponente de la Fuerza Magnética del Ejemplo 2 (pág 9)**
> En el Ejemplo 2 (un haz de protones moviéndose a $v = 3 \times 10^5\text{ m/s}$ en un campo $B = 2\text{ T}$ a un ángulo de $30^\circ$), la fórmula final del libro escribe:
> $$F_B = 4.8 \times 10^{14}\text{ N}$$
> * **Corrección Física:** Esto es un error tipográfico en el exponente. La fuerza real en el protón es infinitesimalmente pequeña, del orden de nano-micro-fuerzas. El valor correcto es:
>   $$F_B = (1.6 \times 10^{-19}\text{ C}) \cdot (3 \times 10^5\text{ m/s}) \cdot (2\text{ T}) \cdot \sin(30^\circ) = 4.8 \times 10^{-14}\text{ N}$$
>   La discrepancia representa 28 órdenes de magnitud.

> [!IMPORTANT]
> **Errata 2: Confusión de Nombre del Ejemplo 2 (pág 9)**
> La conclusión del Ejemplo 2 en el libro dice: "...la fuerza que percibe el **electrón** dentro del cinescopio es de...".
> * **Corrección Física:** El enunciado de este ejercicio define formalmente un haz de **protones**. Por tanto, la conclusión debe referirse al **protón**, no al electrón.


## 3. Cargas en Movimiento en Campos Magnéticos

### 3.1 Movimiento Circular Uniforme
Dado que la fuerza magnética es perpendicular a la velocidad, su producto escalar con el desplazamiento diferencial es cero, por lo que no realiza trabajo y la rapidez de la partícula permanece constante.

Si la velocidad es perpendicular a un campo magnético uniforme, la fuerza magnética actúa como una fuerza centrípeta, forzando a la partícula a describir una órbita circular de radio $r$:
$$F_B = F_c \implies q v B = \frac{m v^2}{r} \implies r = \frac{m v}{q B}$$
El radio de la órbita (radio de ciclotrón) es proporcional al momento lineal e inversamente proporcional al campo magnético y la carga.

La velocidad angular de giro (frecuencia de ciclotrón) y el período de movimiento son:
$$\omega = \frac{v}{r} = \frac{q B}{m}$$
$$T = \frac{2\pi r}{v} = \frac{2\pi}{\omega} = \frac{2\pi m}{q B}$$
Nótese que ni la velocidad angular $\omega$ ni el período $T$ dependen de la velocidad lineal de la partícula ni del radio de la órbita.

### 3.2 Movimiento Helicoidal (Espiral)
Si la velocidad de la partícula tiene una componente paralela al campo ($v_\parallel$) y una componente perpendicular ($v_\perp$):
* El movimiento perpendicular al campo sigue siendo circular uniforme con radio $r = m v_\perp / qB$.
* El movimiento paralelo no experimenta fuerza alguna, por lo que se desplaza a velocidad constante $v_\parallel$.

La superposición de ambos movimientos da lugar a una trayectoria helicoidal (hélice) a lo largo de las líneas de campo. El paso de hélice (pitch) es la distancia que recorre la partícula paralela al campo en un período:
$$p = v_\parallel T = v_\parallel \frac{2\pi m}{q B}$$

### 3.3 Campos Magnéticos No Uniformes
En un campo magnético que aumenta de intensidad hacia los extremos (botella magnética), la componente perpendicular de la fuerza magnética empuja a la partícula cargada en dirección opuesta a la región de campo fuerte. Esto obliga a la partícula a oscilar de un extremo al otro, logrando su confinamiento espacial.

Este fenómeno explica los **Cinturones de Van Allen** de la Tierra, donde partículas del viento solar quedan atrapadas por el campo terrestre, viajando espiralmente de polo a polo geográfico y chocando en la atmósfera superior para producir las auroras polares.


## 4. Fuerzas y Torques en Corrientes Eléctricas

### 4.1 Fuerza en un Alambre Conductor
Una corriente eléctrica en un conductor es un flujo neto de partículas cargadas. Al aplicar un campo magnético externo, la fuerza experimentada por cada carga de deriva se transmite mecánicamente a los átomos del alambre.

Para un segmento recto de longitud $l$ portando una corriente $I$ en un campo magnético uniforme $\vec{B}$:
$$\vec{F}_B = I \vec{l} \times \vec{B}$$
Donde $\vec{l}$ es un vector que apunta en la dirección de la corriente. La magnitud es:
$$F_B = I l B \sin\phi$$

Si el alambre no es recto o el campo no es uniforme, integramos la fuerza infinitesimal sobre los elementos de línea del alambre:
$$d\vec{F}_B = I d\vec{l} \times \vec{B} \implies \vec{F}_B = I \int_{a}^{b} d\vec{l} \times \vec{B}$$

### 4.2 Par de Torsión (Torque) en Espiras y Bobinas
Consideremos una espira plana de área $A$ portando corriente $I$ inmersa en un campo magnético uniforme $\vec{B}$. Definimos el **momento dipolar magnético** $\vec{\mu}$ de la espira como:
$$\vec{\mu} = I \vec{A}$$
Para una bobina con $N$ espiras de igual área, el momento magnético neto es:
$$\vec{\mu} = N I \vec{A}$$
El vector área $\vec{A}$ es perpendicular al plano de la espira, con sentido dado por el cierre de los dedos de la mano derecha en la dirección de la corriente.

El par de torsión vectorial ejercido por el campo sobre el dipolo es:
$$\vec{\tau} = \vec{\mu} \times \vec{B}$$
La magnitud del torque es:
$$\tau = \mu B \sin\phi$$
Donde $\phi$ es el ángulo entre el momento magnético $\vec{\mu}$ (normal a la espira) y el campo magnético $\vec{B}$. El par tiende a rotar la espira hasta alinear $\vec{\mu}$ en la dirección de $\vec{B}$ (donde el torque es cero).

### 4.3 Energía Potencial Magnética y Corrección de Errata de Signo

> [!IMPORTANT]
> **Corrección de Errata del Signo del Dipolo (pág 19):**
> El texto guía escribe la energía de orientación del dipolo como: $U = \vec{\mu} \cdot \vec{B} = \mu B \cos\phi$. Sin embargo, el signo correcto es negativo:
> $$U = -\vec{\mu} \cdot \vec{B} = -\mu B \cos\phi$$
> * **Justificación Física:** El trabajo mecánico para rotar el dipolo se almacena en el sistema. La alineación espontánea paralela al campo ($\phi = 0^\circ$) corresponde al estado de mínima energía (equilibrio estable, $U_{min} = -\mu B$). La orientación antiparalela ($\phi = 180^\circ$) representa la máxima energía (equilibrio inestable, $U_{max} = +\mu B$). La fórmula del libro daría mínimos de energía en la orientación antiparalela, lo cual viola las leyes de la termodinámica y el equilibrio.


## 5. Efecto Hall

El **Efecto Hall** consiste en la aparición de una diferencia de potencial transversal en un conductor por el que circula una corriente cuando se le somete a un campo magnético perpendicular.

### 5.1 Mecanismo Físico y Campo Hall
Al circular una corriente en el eje $x$ dentro de un conductor de ancho $d$ y espesor $t$, y aplicar un campo magnético en el eje $y$, las cargas libres experimentan una fuerza magnética lateral de Lorentz en el eje $z$ (desviándolas hacia un borde del conductor).

Esta acumulación lateral de carga genera un campo eléctrico transversal (campo Hall, $E_H$) que se opone a la desvición de más cargas. En el equilibrio electrostático, la fuerza eléctrica cancela la fuerza magnética:
$$F_e = F_B \implies q E_H = q v_d B \implies E_H = v_d B$$
Donde $v_d$ es la velocidad de deriva de los portadores.

La diferencia de potencial Hall transversal generada es:
$$\Delta V_H = E_H d = v_d B d$$

Sustituyendo la velocidad de deriva en función de la corriente ($v_d = I / n q A$) y el área transversal ($A = d \cdot t$):
$$\Delta V_H = \frac{I B d}{n q (d \cdot t)} \implies \Delta V_H = \frac{I B}{n q t}$$
Donde $n$ es la densidad de portadores de carga y $q$ es la carga (incluyendo su signo). El signo del voltaje Hall permite determinar de manera directa si los portadores de carga en el material son positivos (huecos) o negativos (electrones libres).


```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración estética de Matplotlib
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12, 'figure.titlesize': 13})

# =======================================================================
# SIMULACIÓN 1: TRAYECTORIA HELICOIDAL 3D DE UN PROTÓN EN CAMPO UNIFORME
# =======================================================================
q_e = 1.602e-19    # Carga elemental (C)
m_p = 1.673e-27    # Masa del protón (kg)
B_z = 0.5          # Campo magnético en Z (T)

omega_c = q_e * B_z / m_p # Frecuencia angular de ciclotrón
R_ciclo = 0.05            # Radio de ciclotrón: 5 cm (0.05 m)
v_perp = R_ciclo * omega_c
v_para = 0.25 * v_perp    # Componente paralela a Z

T_ciclo = 2.0 * np.pi / omega_c
t_grid = np.linspace(0, 3.0 * T_ciclo, 1000)

x_t = R_ciclo * np.sin(omega_c * t_grid)
y_t = R_ciclo * np.cos(omega_c * t_grid)
z_t = v_para * t_grid
p_pitch = v_para * T_ciclo

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot(x_t, y_t, z_t, color='crimson', lw=2.5, label='Trayectoria del Protón')

# Trazar algunas líneas de campo de referencia en azul
for x_ref in [-R_ciclo, 0.0, R_ciclo]:
    for y_ref in [-R_ciclo, 0.0, R_ciclo]:
        ax1.plot([x_ref, x_ref], [y_ref, y_ref], [0, np.max(z_t)], color='royalblue', ls=':', alpha=0.3)

ax1.set_title("Trayectoria Helicoidal 3D de un Protón en Campo Uniforme", fontweight='bold')
ax1.set_xlabel("Eje X (m)")
ax1.set_ylabel("Eje Y (m)")
ax1.set_zlabel("Eje Z (m)")
ax1.grid(True, ls='--', alpha=0.5)

# Rotulación informativa de parámetros de ciclotrón
info_str = f"Radio R: {R_ciclo*100:.1f} cm\nFrecuencia f: {omega_c/(2*np.pi)/1e6:.2f} MHz\nPaso (Pitch) p: {p_pitch*100:.1f} cm"
ax1.text2D(0.05, 0.95, info_str, transform=ax1.transAxes, color='black', fontsize=10, 
           bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))

ax1.legend(frameon=True, loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 2: TORQUE Y ENERGÍA POTENCIAL DE UN DIPOLO EN ROTACIÓN
# =======================================================================
mu_mom = 1.18      # Momento dipolar: 1.18 A.m^2 (Ejemplo 5)
B_ext = 1.20       # Campo magnético externo: 1.20 T

phi_deg = np.linspace(0, 360, 500)
phi_rad = np.deg2rad(phi_deg)

tau_vals = mu_mom * B_ext * np.sin(phi_rad)
U_vals = -mu_mom * B_ext * np.cos(phi_rad)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 8))

# Subplot 1: Torque
ax1.plot(phi_deg, tau_vals, 'crimson', lw=2.5, label=r'Par de torsión $\tau(\phi)$')
ax1.axhline(0, color='black', lw=0.8, ls='--')
ax1.axvline(180, color='gray', lw=0.8, ls=':')
ax1.set_title("Par de Torsión vs. Ángulo de Orientación del Dipolo", fontweight='bold')
ax1.set_xlabel(r"Ángulo $\phi$ (grados)")
ax1.set_ylabel(r"Torque $\tau$ (N·m)")
ax1.grid(True, ls='--', alpha=0.5)
ax1.legend(frameon=True)

# Subplot 2: Energía Potencial Corregida (Mínimo en cero)
ax2.plot(phi_deg, U_vals, 'royalblue', lw=2.5, label=r'Energía Potencial $U(\phi)$')
ax2.axhline(0, color='black', lw=0.8, ls='--')
ax2.scatter(0, -mu_mom*B_ext, color='green', s=90, zorder=5, label=r'Equilibrio Estable (Min $U = -\mu B$)')
ax2.scatter(180, mu_mom*B_ext, color='red', s=90, zorder=5, label=r'Equilibrio Inestable (Max $U = +\mu B$)')
ax2.axvline(180, color='gray', lw=0.8, ls=':')
ax2.set_title("Energía Potencial Magnética Corregida vs. Ángulo de Orientación", fontweight='bold')
ax2.set_xlabel(r"Ángulo $\phi$ (grados)")
ax2.set_ylabel("Energía U (J)")
ax2.grid(True, ls='--', alpha=0.5)
ax2.legend(frameon=True)

plt.tight_layout()
plt.show()
```

```python
# =======================================================================
# SIMULACIÓN 3: SUPERFICIE 3D DEL EFECTO HALL EN COBRE
# =======================================================================
n_cu = 8.46e28     # Densidad de portadores del cobre (m^-3)
q_e_val = 1.602e-19 # Carga del electrón (C)
t_thickness = 0.001 # Espesor del alambre: 0.1 cm (0.001 m)

I_vec = np.linspace(0.5, 10.0, 45)  # Corriente (A)
B_vec = np.linspace(0.1, 2.0, 45)   # Campo Magnético (T)
I_grid, B_grid = np.meshgrid(I_vec, B_vec)

# Voltaje Hall en microvoltios
VH_grid_uV = (I_grid * B_grid) / (n_cu * q_e_val * t_thickness) * 1e6

fig = plt.figure(figsize=(10, 8))
ax3 = fig.add_subplot(111, projection='3d')

surf = ax3.plot_surface(I_grid, B_grid, VH_grid_uV, cmap='viridis', edgecolor='none', alpha=0.75)
fig.colorbar(surf, ax=ax3, shrink=0.5, aspect=6, label=r'Voltaje Hall $\Delta V_H$ ($\mu$V)')

# Marcar el punto experimental del Ejemplo 6: I = 5 A, B = 1.2 T
I_ej6 = 5.0
B_ej6 = 1.2
VH_ej6 = (I_ej6 * B_ej6) / (n_cu * q_e_val * t_thickness) * 1e6
ax3.scatter([I_ej6], [B_ej6], [VH_ej6], color='red', s=110, zorder=10, label=fr'Ejemplo 6: {VH_ej6:.2f} $\mu$V')

ax3.set_title(r"Voltaje Hall $\Delta V_H$ en función de la Corriente y el Campo", fontweight='bold')
ax3.set_xlabel("Corriente I (A)")
ax3.set_ylabel("Campo Magnético B (T)")
ax3.set_zlabel(r"Voltaje Hall $\Delta V_H$ ($\mu$V)")
ax3.legend(frameon=True, loc='upper left')
plt.tight_layout()
plt.show()
```

## 6. Verificación Analítica con SymPy

Utilizaremos SymPy para realizar el producto cruzado tridimensional simbólico que define la fuerza de Lorentz y el torque sobre el dipolo magnético. Validaremos de forma algebraica la ortogonalidad intrínseca de estas relaciones físicas, demostrando que la fuerza magnética es perpendicular a la velocidad ($\vec{F}_B \cdot \vec{v} = 0$) y que el torque es perpendicular al momento magnético ($\vec{\tau} \cdot \vec{\mu} = 0$).


```python
import sympy as sp

# Definir variables simbólicas reales
v_x, v_y, v_z = sp.symbols('v_x v_y v_z', real=True)
B_x, B_y, B_z = sp.symbols('B_x B_y B_z', real=True)
mu_x, mu_y, mu_z = sp.symbols('mu_x mu_y mu_z', real=True)
q_sym = sp.symbols('q', real=True)

# Construir vectores en coordenadas cartesianas mediante matrices
vel_vec = sp.Matrix([v_x, v_y, v_z])
B_field = sp.Matrix([B_x, B_y, B_z])
dip_mom = sp.Matrix([mu_x, mu_y, mu_z])

# 1. Fuerza de Lorentz: F = q * (v x B)
Fuerza_L = q_sym * vel_vec.cross(B_field)

# Comprobar ortogonalidades del vector de fuerza
dot_F_v = sp.simplify(Fuerza_L.dot(vel_vec))
dot_F_B = sp.simplify(Fuerza_L.dot(B_field))

# 2. Torque magnético: tau = mu x B
Torque_M = dip_mom.cross(B_field)

# Comprobar ortogonalidades del vector de torque
dot_T_mu = sp.simplify(Torque_M.dot(dip_mom))
dot_T_B = sp.simplify(Torque_M.dot(B_field))

print("======================================================================")
print("VERIFICACIÓN SIMBÓLICA DE ORTOGONALIDAD CON SYMPY")
print("======================================================================")
print(f"Fuerza de Lorentz (F_B):\n{Fuerza_L}\n")
print(f"Torque magnético (tau):\n{Torque_M}\n")
print(f"Fuerza dot Velocidad (F_B . v):  {dot_F_v} (Esperado: 0)")
print(f"Fuerza dot Campo (F_B . B):      {dot_F_B} (Esperado: 0)")
print(f"Torque dot Dipolo (tau . mu):    {dot_T_mu} (Esperado: 0)")
print(f"Torque dot Campo (tau . B):      {dot_T_B} (Esperado: 0)")

assert dot_F_v == 0 and dot_F_B == 0 and dot_T_mu == 0 and dot_T_B == 0, \
    "Error: ¡La validación algebraica de ortogonalidad vectorial falló!"
print("======================================================================")
print("¡VERIFICACIÓN COMPLETA! Ortogonalidad geométrica 3D demostrada analíticamente.")
print("======================================================================")
```

Las **integrales de superficie y de volumen** extienden los conceptos del cálculo multivariable bidimensional a entidades geométricas incrustadas en el espacio tridimensional $\mathbb{R}^3$. Por su parte, el Teorema de Gauss establece una equivalencia exacta entre el flujo neto a través de una superficie cerrada y la acumulación tridimensional de la divergencia del campo dentro del volumen encerrado. ### 1. Superficies en el Espacio y Cálculo de Áreas

Una superficie bidimensional en $\mathbb{R}^3$ se describe paramétricamente mediante una función vectorial de dos variables independientes $u$ y $v$ sobre una región de parámetros $D \subset \mathbb{R}^2$:

$$\mathbf{r}(u, v) = \langle x(u,v), \ y(u,v), \ z(u,v) \rangle$$

El vector normal fundamental a la superficie se obtiene mediante el producto vectorial de sus vectores de velocidad tangente parciales:

$$\mathbf{N}(u, v) = \mathbf{r}_u \times \mathbf{r}_v = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ \frac{\partial x}{\partial u} & \frac{\partial y}{\partial u} & \frac{\partial z}{\partial u} \\ \frac{\partial x}{\partial v} & \frac{\partial y}{\partial v} & \frac{\partial z}{\partial v} \end{vmatrix}$$

El área de la superficie es la integral doble de la magnitud de este vector normal sobre la región $D$:

$$A(S) = \iint_D \|\mathbf{r}_u \times \mathbf{r}_v\| \, du \, dv$$

#### Aclaración Terminológica (Silla de Montar)
> [!NOTE]
> En el material guía de estudio (Ejemplo 3, pág. 5), se denomina "hiperboloide de una hoja" a la superficie descrita por $z = u^2 - v^2$. Esta denominación es incorrecta. La superficie de la forma $z = x^2 - y^2$ representa geométricamente un **paraboloide hiperbólico** (comúnmente conocido como silla de montar). Un hiperboloide de una hoja responde a la ecuación del tipo $x^2 + y^2 - z^2 = 1$.

Graficaremos el paraboloide elíptico, la esfera y el paraboloide hiperbólico en 3D.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Configuración de estilo visual
plt.style.use('seaborn-v0_8-whitegrid')

# Generar mallas para visualización 3D
u = np.linspace(-1.5, 1.5, 50)
v = np.linspace(-1.5, 1.5, 50)
U, V = np.meshgrid(u, v)

# Silla de montar (Paraboloide Hiperbólico)
Z_silla = U**2 - V**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(U, V, Z_silla, cmap='coolwarm', alpha=0.8, edgecolor='none')

ax.set_title('Paraboloide Hiperbólico (Silla de Montar $z = x^2 - y^2$)')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.tight_layout()
plt.show()
```

### 2. Integración sobre Superficies y Flujo Vectorial

Sea $f(x, y, z)$ una función escalar continua. La integral de superficie de $f$ sobre $S$ es:

$$\iint_S f \, dS = \iint_D f(\mathbf{r}(u, v)) \, \|\mathbf{r}_u \times \mathbf{r}_v\| \, du \, dv$$

Si $\mathbf{F}$ es un campo vectorial continuo, el **flujo** de $\mathbf{F}$ a través de $S$ en la dirección de la normal unitaria $\mathbf{n}$ se define como:

$$\text{Flujo} = \iint_S \mathbf{F} \cdot \mathbf{n} \, dS = \iint_D \mathbf{F}(\mathbf{r}(u, v)) \cdot (\mathbf{r}_u \times \mathbf{r}_v) \, du \, dv$$

#### Corrección de Errata de Texto Guía (Evaluación de Sustitución en el Paraboloide)
> [!IMPORTANT]
> **ERRATA DETECTADA EN EL TEXTO GUÍA:**
> En el Ejemplo 9 (pág. 8), se calcula la integral del campo $f(x, y, z) = x^2 + y^2$ sobre el paraboloide $z = x^2 + y^2$ para $0 \leq z \leq 1$.
> La parametrización conduce a la integral:
> 
> $$\iint_S (x^2+y^2) dS = 2\pi \int_0^1 u^3 \sqrt{1 + 4u^2} du$$
> 
> Para resolverla, el texto realiza la misma sustitución errónea del Ejemplo 1 de la Lección 7 ($w = 1+4u^2 \implies dw = 8u du$), ignorando el término residual $u^2$ de la potencia $u^3$, resultando en la integral incorrecta de $\frac{1}{8} \int \sqrt{w} dw$ y afirmando que el resultado es:
> 
> $$\iint_S (x^2+y^2) dS = \frac{\pi}{6}(5\sqrt{5} - 1) \approx 5.3304 \quad \text{[SUSTITUCIÓN INCORRECTA DEL TEXTO]}$$
> 
> **Resolución Analítica Correcta:**
> El factor $u^3$ debe separarse como $u^2 \cdot u$, de modo que $u \, du = \frac{dw}{8}$, y la $u^2$ residual se sustituye por $\frac{w-1}{4}$:
> 
> $$2\pi \int_0^1 u^3 \sqrt{1+4u^2} du = 2\pi \int_1^5 \left( \frac{w-1}{4} \right) \sqrt{w} \, \frac{dw}{8} = \frac{\pi}{16} \int_1^5 (w^{3/2} - w^{1/2}) dw$$
> 
> $$= \frac{\pi}{16} \left[ \frac{2}{5} w^{5/2} - \frac{2}{3} w^{3/2} \right]_1^5 = \pi \left( \frac{5}{12}\sqrt{5} + \frac{1}{60} \right) \approx 2.97931$$
> 
> Evaluaremos numéricamente por parches de superficie para comprobar la convergencia a $2.97931$.

```python
def integral_superficie_discreta(f, r_param, u_min, u_max, v_min, v_max, Nu, Nv):
    u = np.linspace(u_min, u_max, Nu + 1)
    v = np.linspace(v_min, v_max, Nv + 1)
    du = (u_max - u_min) / Nu
    dv = (v_max - v_min) / Nv
    
    suma = 0.0
    for i in range(Nu):
        for j in range(Nv):
            u_mid = 0.5 * (u[i] + u[i+1])
            v_mid = 0.5 * (v[j] + v[j+1])
            
            pos = r_param(u_mid, v_mid)
            val_f = f(*pos)
            
            # Derivadas por diferencias centrales
            eps = 1e-5
            r_u = (r_param(u_mid + eps, v_mid) - r_param(u_mid - eps, v_mid)) / (2.0 * eps)
            r_v = (r_param(u_mid, v_mid + eps) - r_param(u_mid, v_mid - eps)) / (2.0 * eps)
            
            normal = np.cross(r_u, r_v)
            dS = np.linalg.norm(normal) * du * dv
            suma += val_f * dS
    return suma

def f_campo_escalar(x, y, z):
    return x**2 + y**2

def r_paraboloide(u, v):
    return np.array([u * np.cos(v), u * np.sin(v), u**2])

val_real = 2.97931040369
val_pdf = 5.33041350

print("VERIFICACIÓN DE INTEGRAL DE SUPERFICIE ESCALAR (PARCHES):")
for N in [50, 150]:
    aprox = integral_superficie_discreta(f_campo_escalar, r_paraboloide, 0.0, 1.0, 0.0, 2.0*np.pi, N, N)
    print(f"  Nu=Nv={N:3d} | Aprox: {aprox:.6f} | Error vs Real: {abs(aprox - val_real):.2e} | Error vs Texto: {abs(aprox - val_pdf):.2e}")
```

### 3. Teorema de Stokes

El Teorema de Stokes es la generalización tridimensional del Teorema de Green. Relaciona el flujo del rotacional de un campo vectorial a través de una superficie abierta con la circulación del campo a lo largo de su curva cerrada frontera $\partial S$:

$$\iint_S (\nabla \times \mathbf{F}) \cdot d\mathbf{S} = \oint_{\partial S} \mathbf{F} \cdot d\mathbf{r}$$

#### Corrección de Errata de Texto Guía (Stokes en el Semicírculo)
> [!IMPORTANT]
> **ERRATA DETECTADA EN EL TEXTO GUÍA:**
> En el Ejemplo 11 (pág. 9), se pide evaluar Stokes para $\mathbf{F} = \langle y, -x, z \rangle$ sobre el semicírculo unitario en el plano $xy$. La integral del rotacional sobre la superficie da exactamente $-\pi$.
> Sin embargo, al calcular la integral de línea en la frontera, el texto comete dos errores graves:
> 1. Realiza una evaluación incorrecta de signo en la segunda componente del campo, dando $0$ en la integral del arco circular.
> 2. Omite por completo el segmento recto de frontera que une $(-1, 0)$ con $(1, 0)$ sobre el eje $x$.
> Al dar $0$, contradice el valor de la superficie ($-\pi$), y el texto afirma falsamente que "la integral de línea ... es nula".
> 
> **Resolución Correcta:**
> El contorno de frontera cerrado consta de dos segmentos discretos:
> - **Arco Circular ($C_1$):** $\mathbf{r}_A(t) = \langle \cos t, \sin t, 0 \rangle$ para $t \in [0, \pi]$.
>   $$\int_{C_1} \mathbf{F} \cdot d\mathbf{r} = \int_0^{\pi} \langle \sin t, -\cos t, 0 \rangle \cdot \langle -\sin t, \cos t, 0 \rangle \, dt = \int_0^{\pi} (-\sin^2 t - \cos^2 t) \, dt = -\pi$$
> - **Eje X Recto ($C_2$):** $\mathbf{r}_L(t) = \langle t, 0, 0 \rangle$ para $t \in [-1, 1]$ (de regreso de $-1$ a $1$ para cerrar en sentido antihorario).
>   $$\int_{C_2} \mathbf{F} \cdot d\mathbf{r} = \int_{-1}^1 \langle 0, -t, 0 \rangle \cdot \langle 1, 0, 0 \rangle \, dt = 0$$
> 
> Sumando ambos tramos obtenemos el valor exacto de contorno: $-\pi + 0 = -\pi$, demostrando que se cumple de forma impecable el Teorema de Stokes.

```python
# Simular las integrales del Teorema de Stokes en el Semicírculo
def F_stokes(x, y, z):
    return np.array([y, -x, z])

# Contorno cerrado de dos tramos
N_pts = 1000

# Tramo 1: Arco circular t in [0, pi]
t_a = np.linspace(0.0, np.pi, N_pts + 1)
int_arco = 0.0
for i in range(N_pts):
    t_mid = 0.5 * (t_a[i] + t_a[i+1])
    pos = np.array([np.cos(t_mid), np.sin(t_mid), 0.0])
    r_prime = np.array([-np.sin(t_mid), np.cos(t_mid), 0.0])
    int_arco += np.dot(F_stokes(*pos), r_prime * (np.pi / N_pts))

# Tramo 2: Segmento recto de -1 a 1 en el eje x
t_r = np.linspace(-1.0, 1.0, N_pts + 1)
int_recta = 0.0
for i in range(N_pts):
    t_mid = 0.5 * (t_r[i] + t_r[i+1])
    pos = np.array([t_mid, 0.0, 0.0])
    r_prime = np.array([1.0, 0.0, 0.0])
    int_recta += np.dot(F_stokes(*pos), r_prime * (2.0 / N_pts))

stokes_linea_total = int_arco + int_recta
print("VERIFICACIÓN DE STOKES PLANA:")
print(f"  - Circulación en el Arco Circular:     {int_arco:.6f} J")
print(f"  - Circulación en el Tramo Recto:       {int_recta:.6f} J")
print(f"  - Circulación Contorno Cerrado Total:  {stokes_linea_total:.6f} J (Debe ser -pi = {-np.pi:.6f})")
```

### 4. Teorema de Gauss (de la Divergencia)

El Teorema de Gauss relaciona el flujo saliente de un campo vectorial a través de una superficie cerrada orientada positivamente con la integral triple de la divergencia del campo en el volumen encerrado:

$$\iint_S \mathbf{F} \cdot d\mathbf{S} = \iiint_V (\nabla \cdot \mathbf{F}) \, dV$$

#### Caso de Estudio:
Calcularemos el flujo saliente del campo vectorial $\mathbf{F}(x,y,z) = \langle 2x, 3y, 4z \rangle$ a través del cubo de lado 2 centrado en el origen $V = [-1, 1] \times [-1, 1] \times [-1, 1]$:
1. **Divergencia del Campo:** $\nabla \cdot \mathbf{F} = \frac{\partial (2x)}{\partial x} + \frac{\partial (3y)}{\partial y} + \frac{\partial (4z)}{\partial z} = 2 + 3 + 4 = 9$.
2. **Integral de Volumen:** $\iiint_V 9 \, dV = 9 \times \text{Volumen}(V) = 9 \times 8 = 72.0$.

Comprobaremos que la suma de los flujos de superficie evaluada en las 6 caras planas del cubo da exactamente 72.0.

```python
# Flujo a través de las 6 caras del cubo de Gauss
def F_gauss(x, y, z):
    return np.array([2*x, 3*y, 4*z])

N_grid = 100
vals = np.linspace(-1.0, 1.0, N_grid + 1)
dA_face = 4.0 / (N_grid**2)
flujo_gauss = 0.0

# Caras paralelas a yz (x=1 y x=-1)
for i in range(N_grid):
    y_mid = 0.5 * (vals[i] + vals[i+1])
    for j in range(N_grid):
        z_mid = 0.5 * (vals[j] + vals[j+1])
        flujo_gauss += np.dot(F_gauss(1.0, y_mid, z_mid), np.array([1.0, 0.0, 0.0])) * dA_face
        flujo_gauss += np.dot(F_gauss(-1.0, y_mid, z_mid), np.array([-1.0, 0.0, 0.0])) * dA_face

# Caras paralelas a xz (y=1 y y=-1)
for i in range(N_grid):
    x_mid = 0.5 * (vals[i] + vals[i+1])
    for j in range(N_grid):
        z_mid = 0.5 * (vals[j] + vals[j+1])
        flujo_gauss += np.dot(F_gauss(x_mid, 1.0, z_mid), np.array([0.0, 1.0, 0.0])) * dA_face
        flujo_gauss += np.dot(F_gauss(x_mid, -1.0, z_mid), np.array([0.0, -1.0, 0.0])) * dA_face

# Caras paralelas a xy (z=1 y z=-1)
for i in range(N_grid):
    x_mid = 0.5 * (vals[i] + vals[i+1])
    for j in range(N_grid):
        y_mid = 0.5 * (vals[j] + vals[j+1])
        flujo_gauss += np.dot(F_gauss(x_mid, y_mid, 1.0), np.array([0.0, 0.0, 1.0])) * dA_face
        flujo_gauss += np.dot(F_gauss(x_mid, y_mid, -1.0), np.array([0.0, 0.0, -1.0])) * dA_face

print("VERIFICACIÓN DEL TEOREMA DE GAUSS:")
print(f"  - Flujo neto total en la frontera:   {flujo_gauss:.6f}")
print(f"  - Integral triple de la divergencia: {9.0 * 8.0:.6f} (9 * Volumen = 72)")
```

### Resumen de Conceptos Clave

1. **Normal Fundamental:** El vector normal a una superficie se obtiene mediante $\mathbf{r}_u \times \mathbf{r}_v$, su magnitud proporciona el factor de escala diferencial de área $dS$.
2. **Teorema de Stokes:** Generaliza la equivalencia entre integrales de línea de contorno cerrado y dobles del rotacional en $\mathbb{R}^3$.
3. **Teorema de Gauss (de la Divergencia):** Relaciona el flujo saliente en superficies cerradas con la integral triple de la divergencia espacial interna.

### 1. El Espacio $\mathbb{R}^n$ y Álgebra Vectorial

El espacio métrico $\mathbb{R}^n$ es el conjunto de todas las $n$-tuplas ordenadas de números reales:

$$\mathbb{R}^n = \{(x_1, x_2, \dots, x_n) \mid x_i \in \mathbb{R} \quad \forall i = 1, 2, \dots, n\}$$

Para los elementos de este espacio (llamados **vectores**), definimos las siguientes operaciones algebraicas fundamentales:

- **Adición de vectores:** Dados $\mathbf{u} = (u_1, u_2, \dots, u_n)$ y $\mathbf{v} = (v_1, v_2, \dots, v_n)$ en $\mathbb{R}^n$:
  
  $$\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2, \dots, u_n + v_n)$$
  
- **Multiplicación por un escalar:** Dado $\mathbf{u} \in \mathbb{R}^n$ y $c \in \mathbb{R}$:
  
  $$c\mathbf{u} = (cu_1, cu_2, \dots, cu_n)$$
  
- **Producto punto (o producto escalar interno):** El producto punto de $\mathbf{u}$ y $\mathbf{v}$ es un número real definido como:
  
  $$\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^n u_i v_i = u_1 v_1 + u_2 v_2 + \dots + u_n v_n$$
  
- **Norma (o longitud) de un vector:** Representa la distancia euclidiana desde el origen al punto definido por el vector:
  
  $$\|\mathbf{u}\| = \sqrt{\mathbf{u} \cdot \mathbf{u}} = \sqrt{u_1^2 + u_2^2 + \dots + u_n^2}$$

A continuación, implementaremos código para visualizar la adición vectorial en 2D y 3D utilizando flechas en Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D

# Configuración de estilo visual
plt.style.use('seaborn-v0_8-whitegrid')

# Visualización de la suma vectorial en 2D
u = np.array([2, 3])
v = np.array([4, -1])
w = u + v

fig, ax = plt.subplots(figsize=(6, 6))
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{u} = (2, 3)$')
ax.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\mathbf{v} = (4, -1)$')
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\mathbf{u}+\mathbf{v} = (6, 2)$')

ax.set_xlim(-1, 8)
ax.set_ylim(-2, 5)
ax.set_xlabel('Componente X')
ax.set_ylabel('Componente Y')
ax.set_title('Adición Vectorial en 2D')
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 2. Distancias y Sucesiones en $\mathbb{R}^n$

#### Distancia Euclidiana:
La distancia geométrica entre dos puntos $\mathbf{a} = (a_1, a_2, \dots, a_n)$ y $\mathbf{b} = (b_1, b_2, \dots, b_n)$ en $\mathbb{R}^n$ se define a partir de la norma de su diferencia:

$$d(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\| = \sqrt{\sum_{i=1}^n (a_i - b_i)^2}$$

Esta métrica cumple rigurosamente con las propiedades de un espacio métrico:
1. **No negatividad:** $d(\mathbf{a}, \mathbf{b}) \geq 0$, y $d(\mathbf{a}, \mathbf{b}) = 0 \iff \mathbf{a} = \mathbf{b}$.
2. **Simetría:** $d(\mathbf{a}, \mathbf{b}) = d(\mathbf{b}, \mathbf{a})$.
3. **Desigualdad triangular:** Para cualquier tercer punto $\mathbf{c}$:
   
   $$d(\mathbf{a}, \mathbf{b}) \leq d(\mathbf{a}, \mathbf{c}) + d(\mathbf{c}, \mathbf{b})$$

#### Sucesiones en $\mathbb{R}^n$:
Una sucesión de vectores en $\mathbb{R}^n$ es una lista ordenada de vectores $\{\mathbf{a}_k\}_{k=1}^{\infty}$. Se dice que la sucesión **converge** a un punto límite $\mathbf{L} \in \mathbb{R}^n$ si:

$$\lim_{k \to \infty} d(\mathbf{a}_k, \mathbf{L}) = 0$$

Es decir, para cada $\epsilon > 0$, existe un entero $N \in \mathbb{N}$ tal que si $k \geq N$, entonces $d(\mathbf{a}_k, \mathbf{L}) < \epsilon$.

Estudiaremos didácticamente la sucesión de vectores en el plano 

$$\mathbf{a}_k = \left(\frac{1}{k}, \frac{1}{k^2}\right)$$

y demostraremos visualmente su convergencia hacia el límite $\mathbf{L} = (0, 0)$.

```python
# Definición de la sucesión a_k = (1/k, 1/k^2)
k_vals = np.arange(1, 30)
ax_coords = 1.0 / k_vals
ay_coords = 1.0 / (k_vals**2)

# Graficar la trayectoria hacia el origen
plt.figure(figsize=(7, 5))
plt.plot(ax_coords, ay_coords, 'ro--', alpha=0.6, label=r'Términos $\mathbf{a}_k$')
plt.scatter(0, 0, color='green', s=100, zorder=5, label='Límite L=(0,0)')

# Agregar anotaciones para los primeros términos
for i in range(4):
    plt.annotate(f'a_{i+1}', (ax_coords[i], ay_coords[i]), textcoords="offset points", xytext=(5,5), ha='center', fontsize=9)

plt.title('Convergencia de la Sucesión Vectorial en $\mathbb{R}^2$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.legend(frameon=True)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 3. Conceptos Básicos de Topología en $\mathbb{R}^n$

La topología nos permite dar una definición formal al concepto de "cercanía" mediante conjuntos de puntos. Los pilares conceptuales son:

- **Bola abierta:** La bola abierta de radio $\epsilon > 0$ centrada en un punto $\mathbf{x}_0 \in \mathbb{R}^n$ se define como el conjunto de todos los puntos cuya distancia a $\mathbf{x}_0$ es estrictamente menor que $\epsilon$:
  
  $$B_\epsilon(\mathbf{x}_0) = \{\mathbf{y} \in \mathbb{R}^n \mid d(\mathbf{y}, \mathbf{x}_0) < \epsilon\}$$
  
- **Conjunto abierto:** Un conjunto $U \subset \mathbb{R}^n$ es **abierto** si para cada punto $\mathbf{x} \in U$, existe un radio $\epsilon > 0$ tal que la bola abierta $B_\epsilon(\mathbf{x})$ está completamente contenida en $U$:
  
  $$\forall \mathbf{x} \in U, \quad \exists \epsilon > 0 \quad \text{tal que} \quad B_\epsilon(\mathbf{x}) \subseteq U$$
  
- **Conjunto cerrado:** Un conjunto es **cerrado** si su complemento en $\mathbb{R}^n$ es un conjunto abierto. De manera equivalente, un conjunto es cerrado si contiene a todos sus puntos frontera.

#### Demostración Práctica: El Disco Unitario Abierto es Abierto
Consideremos el conjunto $U = \{(x, y) \in \mathbb{R}^2 \mid x^2 + y^2 < 1\}$ (el interior del círculo de radio $1$). Queremos demostrar rigurosamente que $U$ es abierto.

**Prueba:** Sea $\mathbf{p} = (a, b) \in U$. Esto implica que su distancia al origen es menor que $1$: $a^2 + b^2 < 1$.
Definimos el radio $\delta = 1 - \sqrt{a^2 + b^2}$. Como $\mathbf{p} \in U$, se tiene que $\sqrt{a^2 + b^2} < 1$, por lo tanto, $\delta > 0$.
Tomemos un punto cualquiera $\mathbf{q} = (x, y)$ dentro de la bola abierta $B_\delta(\mathbf{p})$, es decir, tal que $d(\mathbf{q}, \mathbf{p}) < \delta$.
Aplicando la desigualdad triangular para la distancia al origen:

$$\|\mathbf{q}\| = \|\mathbf{q} - \mathbf{p} + \mathbf{p}\| \leq \|\mathbf{q} - \mathbf{p}\| + \|\mathbf{p}\| < \delta + \|\mathbf{p}\| = (1 - \|\mathbf{p}\|) + \|\mathbf{p}\| = 1$$

Como $\|\mathbf{q}\| < 1$, el punto $\mathbf{q}$ pertenece a $U$. Esto demuestra que la bola abierta $B_\delta(\mathbf{p})$ está totalmente contenida en $U$, lo que concluye la demostración de que $U$ es abierto.

### 4. Campos Escalares y Vectoriales en la Física

Un campo es una función matemática que asigna un valor a cada punto del espacio.

- **Campo Escalar:** Asigna un número real a cada punto del espacio. Ejemplo: La distribución de temperaturas $T(x, y)$ en una placa de metal:
  
  $$T(x, y) = 100 - 5x^2 - 5y^2$$
  
- **Campo Vectorial:** Asigna un vector a cada punto del espacio. Ejemplos físicos importantes:
  1. **Velocidad de un fluido en rotación (vórtice):**
     $$\mathbf{v}(x, y) = (-y, x)$$
  2. **Campo eléctrico de una carga puntual en el origen:**
     $$\mathbf{E}(x, y) = \frac{k_e q}{r^3} (x, y), \quad r = \sqrt{x^2 + y^2}$$
  3. **Campo magnético alrededor de un hilo conductor con corriente $I$:**
     $$\mathbf{B}(x, y) = \frac{\mu_0 I}{2\pi r^2} (-y, x)$$

Implementaremos gráficos para visualizar estos campos.

```python
# 1. Visualización de Campo Escalar de Temperatura
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
T = 100 - 5*X**2 - 5*Y**2

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Contornos
cp = axs[0].contourf(X, Y, T, cmap='hot', levels=15)
fig.colorbar(cp, ax=axs[0], label='Temperatura (°C)')
axs[0].set_title('Contornos del Campo Térmico $T(x,y)$')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Campo vectorial de velocidad de un fluido v(x, y) = (-y, x)
x_v = np.linspace(-4, 4, 15)
y_v = np.linspace(-4, 4, 15)
X_v, Y_v = np.meshgrid(x_v, y_v)
VX = -Y_v
VY = X_v
axs[1].quiver(X_v, Y_v, VX, VY, color='blue', alpha=0.8)
axs[1].set_title('Campo Vectorial de Velocidad de Fluido')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.tight_layout()
plt.show()

# 2. Graficación del Campo Eléctrico Radial (Carga Puntual)
x_e = np.linspace(-2, 2, 11)
y_e = np.linspace(-2, 2, 11)
X_e, Y_e = np.meshgrid(x_e, y_e)
R = np.hypot(X_e, Y_e)
# Evitar división por cero en el origen
R[R == 0] = np.nan
EX = X_e / R**3
EY = Y_e / R**3

plt.figure(figsize=(6, 5))
plt.quiver(X_e, Y_e, EX, EY, color='red', pivot='middle', label='Campo E')
plt.title('Campo Eléctrico de una Carga Puntual Radial')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 5. Límites, Límites Direccionales y Continuidad

El límite de una función $f(x, y)$ cuando el punto $(x, y)$ se aproxima a $(a, b)$ es igual a $L$:

$$\lim_{(x, y) \to (a, b)} f(x, y) = L$$

si para cada $\epsilon > 0$, existe un $\delta > 0$ tal que si $0 < \sqrt{(x-a)^2 + (y-b)^2} < \delta$, entonces $|f(x, y) - L| < \epsilon$.

#### Límites Direccionales y No Existencia:
Para que un límite bidimensional exista, la función debe aproximarse al mismo valor $L$ independientemente del camino que se tome para llegar a $(a, b)$. Si al evaluar el límite por dos caminos distintos obtenemos valores diferentes, el límite **no existe**.

#### Ejemplo de Análisis: 
Evaluar el límite en el origen de la función:

$$f(x, y) = \frac{3xy}{x^2 + y^2}$$

Aproximémonos a $(0, 0)$ a lo largo del haz de rectas con pendiente $m$ definidas por $y = mx$:

$$f(x, mx) = \frac{3x(mx)}{x^2 + (mx)^2} = \frac{3mx^2}{x^2(1 + m^2)} = \frac{3m}{1+m^2}$$

Tomando el límite cuando $x \to 0$:

$$\lim_{x \to 0} f(x, mx) = \frac{3m}{1+m^2}$$

El valor del límite depende directamente de la pendiente $m$ del camino tomado (ej. si $m=1$, el límite es $3/2$; si $m=0$, el límite es $0$). Por lo tanto, el límite bidimensional no existe.

A continuación, evaluaremos computacionalmente los valores de la función a lo largo de diferentes caminos radiales.

```python
# Evaluar la función f(x, y) = 3xy / (x^2 + y^2) en diferentes trayectorias radiales y = mx
x_vals = np.linspace(-2, 2, 100)
x_vals = x_vals[x_vals != 0] # Evitar el origen

plt.figure(figsize=(8, 5))

for m in [0, 0.5, 1.0, 2.0]:
    y_vals = m * x_vals
    f_vals = (3 * x_vals * y_vals) / (x_vals**2 + y_vals**2)
    plt.plot(x_vals, f_vals, label=f'Trayectoria y = {m}x (Límite: {3*m/(1+m**2):.2f})')

plt.title('Evaluación de Límites Direccionales en el Origen')
plt.xlabel('x')
plt.ylabel('Valor de $f(x, y)$')
plt.legend(frameon=True)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 6. Curvas en el Espacio y Parametrizaciones

Una curva en el espacio tridimensional puede describirse mediante una función con valores vectoriales (o ecuación paramétrica vectorial):

$$\mathbf{r}(t) = \langle x(t), y(t), z(t) \rangle, \quad t \in I$$

La función es continua si cada una de sus componentes $x(t)$, $y(t)$ y $z(t)$ son funciones reales continuas.

#### Ejemplos Clásicos:
1. **Hélice Circular:** Se enrolla sobre un cilindro vertical de radio $1$:
   
   $$\mathbf{r}(t) = \langle \cos t, \sin t, t \rangle$$
   
2. **Curva de Intersección de Superficies:** La intersección del cilindro vertical $x^2 + y^2 = 1$ y del plano inclinado $y + z = 2$ se parametriza como:
   
   $$\mathbf{r}(t) = \langle \cos t, \sin t, 2 - \sin t \rangle$$

Visualizaremos estas dos curvas en el espacio tridimensional.

```python
t = np.linspace(0, 4*np.pi, 200)

fig = plt.figure(figsize=(12, 5))

# Hélice circular
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(np.cos(t), np.sin(t), t, 'b-', linewidth=2.5, label='Hélice')
ax1.set_title('Hélice Circular $\mathbf{r}(t) = \langle \cos t, \sin t, t \rangle$')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend()

# Intersección cilindro-plano
t_int = np.linspace(0, 2*np.pi, 100)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot(np.cos(t_int), np.sin(t_int), 2 - np.sin(t_int), 'r-', linewidth=3, label='Intersección')
ax2.set_title('Intersección Cilindro-Plano')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()

plt.tight_layout()
plt.show()
```

### 7. Geometría Diferencial: Triedro de Frenet-Serret y Errata del Libro

Para una curva paramétrica regular y diferenciable $\mathbf{r}(t)$, definimos las siguientes propiedades geométricas locales:

- **Longitud de arco:** $L = \int_a^b \|\mathbf{r}'(t)\| dt$.
- **Triedro de Frenet-Serret (T, N, B):** Consiste en tres vectores ortonormales móviles que describen la orientación local de la curva:
  - **Tangente unitario:** $\mathbf{T}(t) = \frac{\mathbf{r}'(t)}{\|\mathbf{r}'(t)\|}$
  - **Normal unitario:** $\mathbf{N}(t) = \frac{\mathbf{T}'(t)}{\|\mathbf{T}'(t)\|}$ (apunta al centro de curvatura)
  - **Binormal unitario:** $\mathbf{B}(t) = \mathbf{T}(t) \times \mathbf{N}(t)$
- **Curvatura ($\kappa$):** Mide la tasa de cambio de la dirección de la curva:
  
  $$\kappa = \frac{\|\mathbf{r}'(t) \times \mathbf{r}''(t)\|}{\|\mathbf{r}'(t)\|^3}$$
  
- **Torsión ($\tau$):** Mide la tendencia de la curva a salirse del plano osculador (definido por $\mathbf{T}$ y $\mathbf{N}$):
  
  $$\tau = \frac{(\mathbf{r}'(t) \times \mathbf{r}''(t)) \cdot \mathbf{r}'''(t)}{\|\mathbf{r}'(t) \times \mathbf{r}''(t)\|^2}$$

#### Corrección de la Errata en la Literatura (Torsión de la Hélice)
> [!IMPORTANT]
> **ERRATA DETECTADA EN EL TEXTO GUÍA:**
> En el material de estudio (pág. 14), se define erróneamente el vector binormal de la hélice $\mathbf{r}(t) = \langle \cos t, \sin t, t \rangle$ como:
> 
> $$\mathbf{B}(t) = \left( \frac{\cos t}{\sqrt{2}}, \frac{\sin t}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) \quad \text{[FÓRMULA INCORRECTA DEL TEXTO]}$$
> 
> y a partir de allí calcula su derivada como $\mathbf{B}'(t) = \left( -\frac{\sin t}{\sqrt{2}}, \frac{\cos t}{\sqrt{2}}, 0 \right)$. El texto afirma que al hacer el producto punto con $\mathbf{N}(t) = \langle -\cos t, -\sin t, 0 \rangle$, el resultado da $\frac{\sin^2 t + \cos^2 t}{2} = 1/2$. 
> 
> Sin embargo, esto es un **error algebraico grave**: el producto escalar real de esos dos vectores es:
> 
> $$\mathbf{B}'_{texto}(t) \cdot \mathbf{N}(t) = \left( -\frac{\sin t}{\sqrt{2}} \right)(-\cos t) + \left( \frac{\cos t}{\sqrt{2}} \right)(-\sin t) = \frac{\sin t \cos t - \sin t \cos t}{\sqrt{2}} = 0 \quad \text{[ERROR ALGEBRAICO EN EL TEXTO]}$$
> 
> **DEDUCCIÓN FORMAL Y CORRECTA:**
> El vector binormal real se obtiene a través del producto vectorial $\mathbf{B} = \mathbf{T} \times \mathbf{N}$:
> 
> $$\mathbf{T}(t) = \frac{1}{\sqrt{2}} \langle -\sin t, \cos t, 1 \rangle, \quad \mathbf{N}(t) = \langle -\cos t, -\sin t, 0 \rangle$$
> 
> $$\mathbf{B}(t) = \mathbf{T} \times \mathbf{N} = \frac{1}{\sqrt{2}} \langle \sin t, -\cos t, 1 \rangle \quad \text{[VALOR REAL CORRECTO]}$$
> 
> Derivando el binormal correcto respecto de $t$:
> 
> $$\mathbf{B}'(t) = \frac{1}{\sqrt{2}} \langle \cos t, \sin t, 0 \rangle$$
> 
> Ahora calculamos el producto escalar con $\mathbf{N}(t)$:
> 
> $$\mathbf{B}'(t) \cdot \mathbf{N}(t) = \frac{1}{\sqrt{2}} \langle \cos t, \sin t, 0 \rangle \cdot \langle -\cos t, -\sin t, 0 \rangle = -\frac{\cos^2 t + \sin^2 t}{\sqrt{2}} = -\frac{1}{\sqrt{2}}$$
> 
> Finalmente, calculamos la torsión $\tau(t)$:
> 
> =$$ \tau(t) = -\frac{\mathbf{B}'(t) \cdot \mathbf{N}(t)}{\|\mathbf{r}'(t)\|} = -\frac{-1/\sqrt{2}}{\sqrt{2}} = \frac{1}{2} \quad \text{[DEMOSTRACIÓN COMPLETA Y CORRECTA]}$$

A continuación, utilizaremos el motor de cálculo simbólico de `SymPy` para verificar analíticamente los vectores unitarios, la curvatura y la torsión de la hélice.

```python
# Definición simbólica
t_sym = sp.Symbol('t', real=True)
r_sym = sp.Matrix([sp.cos(t_sym), sp.sin(t_sym), t_sym])

# Velocidad y su norma
v_sym = r_sym.diff(t_sym)
v_norm = v_sym.norm()

# Triedro de Frenet-Serret
T_sym = sp.simplify(v_sym / v_norm)
dT_dt = T_sym.diff(t_sym)
N_sym = sp.simplify(dT_dt / dT_dt.norm())
B_sym = sp.simplify(T_sym.cross(N_sym))

# Curvatura y Torsión
kappa_sym = sp.simplify(dT_dt.norm() / v_norm)
dB_dt = B_sym.diff(t_sym)
tau_sym = sp.simplify(- (dB_dt.dot(N_sym)) / v_norm)

print("--- RESULTADOS SIMBÓLICOS EXACTOS CON SYMPY ---")
print("Vector Tangente T(t):")
sp.pprint(T_sym)
print("\nVector Normal N(t):")
sp.pprint(N_sym)
print("\nVector Binormal B(t) (Correcto):")
sp.pprint(B_sym)
print(f"\nCurvatura kappa: {kappa_sym}")
print(f"Torsión tau:      {tau_sym}")

# Verificación analítica del producto escalar del PDF que da 0 en realidad
N_pdf = sp.Matrix([-sp.cos(t_sym), -sp.sin(t_sym), 0])
dB_pdf = sp.Matrix([-sp.sin(t_sym)/sp.sqrt(2), sp.cos(t_sym)/sp.sqrt(2), 0])
prod_punto_pdf = sp.simplify(dB_pdf.dot(N_pdf))
print(f"\nVerificación de la errata: Producto punto del texto es igual a {prod_punto_pdf} (en lugar del 1/2 reportado)")
```

### Resumen de Conceptos Clave

1. **Espacio Vectorial $\mathbb{R}^n$:** Estructura geométrica dotada de operaciones internas (suma) y externas (producto por escalar), así como de producto interno que induce longitud (norma) y ángulo.
2. **Espacio Métrico y Sucesiones:** La distancia euclidiana cuantifica la separación de puntos y permite analizar la convergencia de trayectorias vectoriales al infinito.
3. **Topología:** El concepto de bola abierta fundamenta la definición rigurosa de límites multidimensionales, diferenciando conjuntos abiertos y cerrados.
4. **Campos en Física:** Los campos escalares asignan temperatura o potencial, mientras que los campos vectoriales describen fuerzas eléctricas y magnéticas espaciales.
5. **Triedro de Frenet-Serret:** Marco de referencia ortonormal móvil ($\mathbf{T}$, $\mathbf{N}$, $\mathbf{B}$) que define la geometría intrínseca de una curva, caracterizada por la curvatura $\kappa$ (flexión) y la torsión $\tau$ (alabeo).
